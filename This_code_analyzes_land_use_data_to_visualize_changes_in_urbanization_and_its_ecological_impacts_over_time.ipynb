This notebook analyzes land use and cover changes in Lahore from 2014 to 2024.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for LULC changes
data = {
    'Year': [2014, 2024],
    'Water Bodies': [30.24, 19.19],
    'Vegetation': [899.73, 812.29],
    'Barren Land': [131.45, 76.58],
    'Build-up': [815.12, 968.5]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
for column in df.columns[1:]:
    plt.plot(df['Year'], df[column], marker='o', label=column)
plt.title('LULC Changes in Lahore (2014-2024)')
plt.xlabel('Year')
plt.ylabel('Area (km²)')
plt.legend()
plt.grid()
plt.show()
```

This analysis provides insights into the trends of land use changes in Lahore.
