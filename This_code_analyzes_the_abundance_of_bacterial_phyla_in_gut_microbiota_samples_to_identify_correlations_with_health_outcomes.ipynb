### Analysis of Bacterial Phyla Abundance in Gut Microbiota

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for bacterial phyla abundance
data = {'Phylum': ['Proteobacteria', 'Actinobacteria', 'Firmicutes', 'Bacteroidetes'], 'Abundance': [20, 10, 40, 30]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Phylum'], df['Abundance'], color=['blue', 'orange', 'green', 'red'])
plt.title('Relative Abundance of Bacterial Phyla in Gut Microbiota')
plt.xlabel('Bacterial Phylum')
plt.ylabel('Abundance (%)')
plt.show()
```
