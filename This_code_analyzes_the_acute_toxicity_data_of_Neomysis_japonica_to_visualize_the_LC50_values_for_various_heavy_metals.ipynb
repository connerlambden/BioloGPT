# Acute Toxicity Analysis of Neomysis japonica
This notebook analyzes the acute toxicity data of Neomysis japonica to visualize the LC50 values for various heavy metals.

```python
import pandas as pd
import plotly.express as px

# Data for LC50 values
data = {'Heavy Metal': ['Copper', 'Cadmium', 'Zinc', 'Arsenic', 'Chromium', 'Lead'],
        'LC50 (µg/L)': [82.3, 90.4, 646.1, 1488.5, 1609.6, 1758.8]}
df = pd.DataFrame(data)

# Create bar plot
fig = px.bar(df, x='Heavy Metal', y='LC50 (µg/L)', title='96 hr LC50 Values for Neomysis japonica')
fig.show()
```

The bar plot above illustrates the acute toxicity of various heavy metals on Neomysis japonica, highlighting the most toxic metals.
