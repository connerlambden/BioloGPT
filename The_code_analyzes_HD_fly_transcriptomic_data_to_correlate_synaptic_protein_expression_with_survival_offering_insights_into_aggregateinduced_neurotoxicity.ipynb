This notebook processes transcriptomic data and survival metrics from HD fly models, specifically focusing on synaptic protein genes including Vmat.

```python
import pandas as pd
import plotly.express as px
# Load the dataset from HD fly experiments
data = pd.read_csv('HD_transgenic_fly_data.csv')
# Calculate correlation between Vmat expression and survival
correlation = data['Vmat_expression'].corr(data['survival_days'])
print('Vmat-Survival Correlation:', correlation)
# Create a scatter plot
fig = px.scatter(data, x='Vmat_expression', y='survival_days', title='Correlation of Vmat Expression with Survival')
fig.show()
```

The analysis also includes multivariate regression to adjust for other synaptic proteins affecting survival.

```python
import statsmodels.api as sm
X = data[['Vmat_expression', 'Other_synaptic_protein']]
X = sm.add_constant(X)
y = data['survival_days']
model = sm.OLS(y, X).fit()
print(model.summary())
```
