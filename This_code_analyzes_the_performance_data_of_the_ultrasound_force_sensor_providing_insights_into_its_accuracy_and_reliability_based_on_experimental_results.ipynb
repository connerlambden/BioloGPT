### Performance Analysis of Ultrasound Force Sensor

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for analysis
weights = [0, 4.907, 19.628, 34.349]
errors = [0.006, 0.015, 0.18, 0.14]

# Create DataFrame
data = pd.DataFrame({'Weight (N)': weights, 'Average Error (N)': errors})

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(data['Weight (N)'], data['Average Error (N)'], color='blue')
plt.title('Average Error of Force Sensor')
plt.xlabel('Applied Weight (N)')
plt.ylabel('Average Error (N)')
plt.grid(True)
plt.show()
```

### Discussion
This analysis provides a visual representation of the average error associated with different applied weights, highlighting the sensor's performance.
