The notebook describes the process of loading 16S rRNA amplicon data, performing quality control, and calculating Shannon diversity indices across different feeding regimes using QIIME2 and relevant Python libraries.

```python
import qiime2
import pandas as pd
import numpy as np
from skbio.diversity.alpha import shannon
# Load dataset
metadata = pd.read_csv('metadata.csv')
data = pd.read_csv('feature-table.csv')
# Calculate Shannon diversity
metadata['shannon'] = data.apply(lambda row: shannon(row.values), axis=1)
print(metadata[['sample_id', 'shannon']].head())
```

This code is useful for analyzing how different feeding parameters affect gut microbiome diversity by directly computing Shannon diversity from microbial abundances.

```python
# The following section plots the diversity data
import plotly.express as px
fig = px.bar(metadata, x='sample_id', y='shannon', title='Shannon Diversity across Samples')
fig.show()
```
