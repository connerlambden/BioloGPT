This notebook imports chemical composition and activity data for basil oils and uses Plotly to generate comparative graphs.

```python
import plotly.graph_objects as go

species = ['O. basilicum', 'O. tenuiflorum', 'O. gratissimum']
growth_inhibition = [2.0, 1.5, 2.5]
afb1_inhibition = [1.25, 0.75, 1.0]

fig = go.Figure(data=[
    go.Bar(name='Growth Inhibition (µL/mL)', x=species, y=growth_inhibition),
    go.Bar(name='AFB1 Inhibition (µL/mL)', x=species, y=afb1_inhibition)
])

fig.update_layout(barmode='group', title='Antifungal Activity of Basil Essential Oils', xaxis_title='Basil Species', yaxis_title='Inhibitory Concentration (µL/mL)')
fig.show()
```

The plot helps visualize differential efficacy among the basil species.
