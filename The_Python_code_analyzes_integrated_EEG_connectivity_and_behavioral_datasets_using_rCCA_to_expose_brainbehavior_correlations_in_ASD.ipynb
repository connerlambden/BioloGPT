This notebook downloads and processes EEG connectivity and behavioral data, applying rCCA to dissect brain-behavior relationships in ASD.

```python
import pandas as pd
import numpy as np
from sklearn.cross_decomposition import CCA
# Load EEG connectivity and behavioral data
connectivity_data = pd.read_csv('eeg_connectivity.csv')
behavior_data = pd.read_csv('behavior_data.csv')
cca = CCA(n_components=3)
X_c, Y_c = cca.fit_transform(connectivity_data, behavior_data)
print('Canonical correlations:', np.corrcoef(X_c.T, Y_c.T))
```

The code then visualizes the first canonical correlation to help elucidate the relationships between connectivity and behavior.

```python
import matplotlib.pyplot as plt
plt.scatter(X_c[:,0], Y_c[:,0], color='#6A0C76')
plt.title('Canonical Correlation Plot')
plt.xlabel('Connectivity Canonical Variate 1')
plt.ylabel('Behavioral Canonical Variate 1')
plt.show()
```
