This notebook will analyze RNA-seq data from KAT6B overexpressing and control samples.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the RNA-seq data
rna_seq_data = pd.read_csv('path_to_rna_seq_data.csv')

# Filter for Tbx1 and Tbx15
filtered_data = rna_seq_data[rna_seq_data['gene'].isin(['Tbx1', 'Tbx15'])]

# Plot the expression levels
sns.boxplot(x='gene', y='expression', data=filtered_data)
plt.title('Expression Levels of Tbx1 and Tbx15')
plt.show()
```

The results will provide insights into the expression dynamics of Tbx1 and Tbx15 under KAT6B overexpression.
