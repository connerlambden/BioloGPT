### Climate Data Analysis for Dengue Prediction
This notebook will analyze historical climate data alongside dengue outbreak records to identify predictive thresholds.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load climate and dengue data
data = pd.read_csv('climate_dengue_data.csv')

# Analyze correlations
correlation = data.corr()
print(correlation)

# Visualize the data
plt.scatter(data['Temperature'], data['Dengue_Cases'])
plt.title('Temperature vs Dengue Cases')
plt.xlabel('Temperature (°C)')
plt.ylabel('Dengue Cases')
plt.show()
```
