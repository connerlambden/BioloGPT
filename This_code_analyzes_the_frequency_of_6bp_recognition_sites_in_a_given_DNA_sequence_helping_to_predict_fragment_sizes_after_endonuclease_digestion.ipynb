### DNA Fragmentation Analysis
This notebook analyzes a DNA sequence to identify potential 6-bp recognition sites for endonucleases.

```python
# Sample DNA sequence
sequence = 'ATCGGAATTCGATCGGATCGAATTCGATCGGATCGGAATTC'

# Function to find 6-bp recognition sites
def find_recognition_sites(seq, site_length=6):
    sites = []
    for i in range(len(seq) - site_length + 1):
        sites.append(seq[i:i + site_length])
    return sites

# Find recognition sites
recognition_sites = find_recognition_sites(sequence)
recognition_sites
```

### Results
The identified recognition sites can be analyzed for their frequency and potential cleavage outcomes.

```python
# Count frequency of each site
from collections import Counter
site_counts = Counter(recognition_sites)
site_counts.most_common()
```
