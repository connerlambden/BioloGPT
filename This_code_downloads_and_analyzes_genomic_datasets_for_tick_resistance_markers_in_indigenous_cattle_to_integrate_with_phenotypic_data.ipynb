This notebook describes the process of downloading genomic data, filtering for candidate genes related to tick resistance, and integrating with phenotypic records from tropical cattle breeds.

```python
import pandas as pd
# Code to load genomic and phenotypic datasets
genomic_data = pd.read_csv('genomic_ticks.csv')
phenotypic_data = pd.read_csv('phenotype_ticks.csv')
# Merge datasets on cattle ID
merged_data = pd.merge(genomic_data, phenotypic_data, on='cattle_id')
# Analysis would include statistical association tests
import statsmodels.api as sm
X = merged_data[['gene_expression_marker1', 'gene_expression_marker2']]
Y = merged_data['tick_burden']
model = sm.OLS(Y, sm.add_constant(X)).fit()
print(model.summary())
```

This step-by-step analysis showcases the integration of genomic markers with phenotypic tick burden data to uncover potential associations that underpin tick-resistance traits.

```python
# Additional code for visualizing the association
import matplotlib.pyplot as plt
plt.scatter(merged_data['gene_expression_marker1'], merged_data['tick_burden'], color='#6A0C76')
plt.xlabel('Gene Expression Marker 1')
plt.ylabel('Tick Burden')
plt.title('Association between Gene Expression and Tick Burden')
plt.show()
```

The final visualization helps in identifying trends and supports further hypothesis testing and refinement.

```python
# Final code cell for saving cleaned data
merged_data.to_csv('cleaned_tick_resistance_data.csv', index=False)
print('Cleaned data saved successfully.')
```
