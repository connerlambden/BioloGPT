This notebook visualizes the relationship between Moringa extract concentration and cyst evagination inhibition, aiding in dose-response analysis.

```python
import plotly.express as px
import pandas as pd

# Define dataset
 data = pd.DataFrame({
    'Concentration': ['0.006 mg/µl', '0.06 mg/µl', '0.3 mg/µl', '0.6 mg/µl'],
    'Evagination Rate': [45, 20, 25, 15]
})

# Create a line plot for the dose-response curve
fig = px.line(data, x='Concentration', y='Evagination Rate', title='Dose-Response Curve for Moringa Extract', markers=True, line_shape='spline', render_mode='svg')
fig.update_traces(line=dict(color='#6A0C76'))
fig.show()
```

The interactive plot helps in clearly understanding how increasing extract concentration relates to decreased cyst evagination.
