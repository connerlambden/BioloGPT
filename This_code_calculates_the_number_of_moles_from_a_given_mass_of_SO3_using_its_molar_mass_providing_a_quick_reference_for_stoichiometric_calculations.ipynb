### Calculate Moles of SO3 from Mass
This notebook calculates the number of moles in a specified mass of SO3.

```python
mass_so3 = 0.5  # in grams
molar_mass_so3 = 80.07  # g/mol
moles_so3 = mass_so3 / molar_mass_so3
moles_so3
```

The calculated number of moles of SO3 is displayed above.

```python
# Displaying the result
print(f'Number of moles in {mass_so3}g of SO3: {moles_so3}')
```
