This notebook imports DEGs and metabolite data, merges them, calculates Pearson correlations, and visualizes significant associations.

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import plotly.express as px

# Load datasets
deg_data = pd.read_csv('deg_data.csv')  # Contains gene expression profiles
metabo_data = pd.read_csv('metabolite_data.csv')  # Contains metabolite intensities

# Merge datasets on a common identifier such as 'sample_id'
merged_data = pd.merge(deg_data, metabo_data, on='sample_id')

# Compute Pearson correlation between gene expression and metabolite levels
merged_data['correlation'] = merged_data.apply(lambda row: pearsonr(row['expression_values'], row['metabolite_levels'])[0], axis=1)

# Filter for significant correlation pairs
significant_pairs = merged_data[merged_data['correlation'] > 0.8]
print(significant_pairs.head())

# Visualize correlations
fig = px.scatter(significant_pairs, x='expression_values', y='metabolite_levels', color='correlation',
                 title='Significant Gene-Metabolite Correlations')
fig.show()
```

The code above outlines the steps to merge omics data and identify candidate gene-metabolite pairs likely involved in CSO biosynthesis.
