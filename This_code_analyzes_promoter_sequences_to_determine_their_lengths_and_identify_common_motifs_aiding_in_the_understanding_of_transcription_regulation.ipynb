### Promoter Length Analysis
This notebook will analyze a dataset of bacterial promoter sequences to determine their lengths and identify common motifs.

```python
import pandas as pd

# Load dataset of promoter sequences
dataset = pd.read_csv('promoter_sequences.csv')

# Calculate lengths of each promoter sequence
dataset['length'] = dataset['sequence'].apply(len)

# Display the average length of promoters
average_length = dataset['length'].mean()
average_length
```

### Results
The average length of the analyzed promoter sequences will provide insights into typical promoter characteristics.

```python
# Display the average length
print(f'Average Promoter Length: {average_length} bp')
```
