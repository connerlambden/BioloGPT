### Soil Enzyme Activity Analysis Based on Magnesium Levels

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Magnesium Level': [5, 10, 15, 20], 'Urease Activity': [0.5, 1.0, 1.5, 2.0], 'Acid Phosphatase Activity': [0.3, 0.6, 0.9, 1.2]}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 5))
plt.plot(df['Magnesium Level'], df['Urease Activity'], marker='o', label='Urease Activity')
plt.plot(df['Magnesium Level'], df['Acid Phosphatase Activity'], marker='o', label='Acid Phosphatase Activity')
plt.title('Soil Enzyme Activities vs. Magnesium Levels')
plt.xlabel('Magnesium Level (mg/kg)')
plt.ylabel('Enzyme Activity (µg/g)')
plt.legend()
plt.grid()
plt.show()
```

### Discussion
This analysis helps visualize the relationship between magnesium levels and soil enzyme activities, indicating potential thresholds for optimal enzyme function.
