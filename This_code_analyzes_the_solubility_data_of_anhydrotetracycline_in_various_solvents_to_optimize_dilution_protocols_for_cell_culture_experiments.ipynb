### Solubility Analysis of Anhydrotetracycline
This notebook will analyze the solubility of anhydrotetracycline in DMSO and ethanol, providing insights for optimal dilution strategies.

```python
import pandas as pd

# Sample solubility data
solubility_data = {
    'Solvent': ['DMSO', 'Ethanol'],
    'Solubility (mg/mL)': [20, 15.43]
}

solubility_df = pd.DataFrame(solubility_data)
solubility_df
```

### Results
The solubility of anhydrotetracycline in different solvents is displayed in the table above.

```python
import matplotlib.pyplot as plt

plt.bar(solubility_df['Solvent'], solubility_df['Solubility (mg/mL)'], color=['#6A0C76', '#FF5733'])
plt.title('Solubility of Anhydrotetracycline in Different Solvents')
plt.ylabel('Solubility (mg/mL)')
plt.show()
```
