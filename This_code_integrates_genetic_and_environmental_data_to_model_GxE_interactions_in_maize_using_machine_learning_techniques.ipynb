### Step 1: Import Libraries
Import necessary libraries for data manipulation and machine learning.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

### Step 2: Load Data
Load genetic and environmental datasets.

```python
# Load datasets
genetic_data = pd.read_csv('genetic_data.csv')
environmental_data = pd.read_csv('environmental_data.csv')
```

### Step 3: Data Preprocessing
Merge datasets and preprocess for modeling.

```python
# Merge datasets
merged_data = pd.merge(genetic_data, environmental_data, on='sample_id')
# Handle missing values
merged_data.fillna(method='ffill', inplace=True)
```
