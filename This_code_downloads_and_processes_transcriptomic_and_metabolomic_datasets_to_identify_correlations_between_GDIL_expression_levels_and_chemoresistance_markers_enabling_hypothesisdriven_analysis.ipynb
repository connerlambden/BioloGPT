Step-by-step analysis: Load raw RNA-seq counts, normalize the data, and run differential expression analysis comparing platinum-resistant vs sensitive samples.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load RNA-seq dataset (replace 'path_to_data' with actual data path)
data = pd.read_csv('path_to_data.csv')

# Normalize and log-transform data
normalized_data = np.log2(data + 1)

# Differential expression analysis
resistant = normalized_data[normalized_data['status'] == 'resistant']
sensitive = normalized_data[normalized_data['status'] == 'sensitive']

diff_expr = resistant.mean() - sensitive.mean()
print(diff_expr[['GDIL', 'CHAC1', 'XRN2']])
```

The analysis continues by correlating expression levels with metabolomic profiles (GSH measurements) to validate GDIL's impact.

```python
import matplotlib.pyplot as plt

# Plot correlation between GDIL expression and GSH levels
plt.figure(figsize=(8,6))
plt.scatter(normalized_data['GDIL'], normalized_data['GSH'], c='teal', label='Samples')
plt.xlabel('GDIL Expression (log2 normalized)')
plt.ylabel('GSH Levels')
plt.title('Correlation between GDIL Expression and Glutathione Levels')
plt.legend()
plt.show()
```

The code then performs statistical tests to determine significance and outputs results for further investigation.

```python
t_stat, p_val = stats.ttest_ind(resistant['GDIL'], sensitive['GDIL'])
print('T-statistic:', t_stat, 'P-value:', p_val)
```
