This notebook will analyze health data from Bhopal survivors, focusing on chronic illnesses and their correlation with exposure levels.

```python
import pandas as pd

# Load health data
health_data = pd.read_csv('bhopal_health_data.csv')

# Analyze chronic illness patterns
chronic_illnesses = health_data[health_data['chronic_illness'] == True]
chronic_illnesses.describe()
```

The analysis will provide insights into the prevalence of chronic illnesses among survivors and their potential links to MIC exposure.

```python
# Visualize the data
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(chronic_illnesses['age'], bins=20, color='blue', alpha=0.7)
plt.title('Age Distribution of Chronic Illnesses in Bhopal Survivors')
plt.xlabel('Age')
plt.ylabel('Number of Cases')
plt.show()
```
