### Genetic Variant Analysis for Sickle Cell Disease
This analysis focuses on identifying variants in the HBB gene that contribute to sickle cell disease and their population frequencies.

```python
import pandas as pd

# Load genetic variant data
variant_data = pd.read_csv('sickle_cell_variants.csv')

# Analyze allele frequencies
allele_frequencies = variant_data.groupby('variant_id')['allele_frequency_exome'].mean()

# Identify high-frequency variants
high_freq_variants = allele_frequencies[allele_frequencies > 0.01]

# Output results
print(high_freq_variants)
```
