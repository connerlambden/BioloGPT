### Enzyme Activity Analysis
This notebook analyzes the relationship between substrate concentration and enzyme activity using Michaelis-Menten kinetics.

```python
import numpy as np
import matplotlib.pyplot as plt

# Substrate concentrations and corresponding enzyme activities
substrate_concentration = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3])
enzyme_activity = np.array([0, 30, 80, 120, 150, 170, 174])

# Plotting the data
plt.plot(substrate_concentration, enzyme_activity, marker='o')
plt.title('Effect of Substrate Concentration on Enzyme Activity')
plt.xlabel('Substrate Concentration (g/L)')
plt.ylabel('Enzyme Activity (U/mL)')
plt.grid()
plt.show()
```

This analysis provides insights into the enzyme kinetics and can be used to optimize conditions for enzymatic reactions.
