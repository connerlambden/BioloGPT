### Step 1: Define the Contig Lengths
We will define a list of contig lengths to calculate the N50 value.

```python
contig_lengths = [500, 1500, 3000, 2000, 1000, 4000, 2500]  # Example contig lengths in bp
```

### Step 2: Calculate the Total Length and N50
We will calculate the total length of the contigs and then determine the N50 value.

```python
total_length = sum(contig_lengths)
contig_lengths.sort(reverse=True)  # Sort in descending order
cumulative_length = 0
n50 = 0
for length in contig_lengths:
    cumulative_length += length
    if cumulative_length >= total_length / 2:
        n50 = length
        break
n50  # This will give the N50 value.
```

### Step 3: Output the N50 Value
The calculated N50 value will be displayed.

```python
print(f'N50 Value: {n50} bp')
```
