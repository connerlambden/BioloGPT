Cette analyse se concentre sur les données d'IRMf pour étudier l'activation des neurones miroirs.

```python
import pandas as pd
# Charger les données d'activation des neurones miroirs
activation_data = pd.read_csv('activation_neurones_miroirs.csv')
```

Nous allons examiner les corrélations entre l'activation des neurones miroirs et les réponses émotionnelles.

```python
correlation = activation_data.corr()  # Calculer les corrélations entre les variables.
```

Cette analyse nous aidera à comprendre comment l'activation des neurones miroirs est liée aux réponses émotionnelles observées.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(correlation, annot=True)
plt.title('Corrélations entre l'activation des neurones miroirs et les réponses émotionnelles')
plt.show()
```
