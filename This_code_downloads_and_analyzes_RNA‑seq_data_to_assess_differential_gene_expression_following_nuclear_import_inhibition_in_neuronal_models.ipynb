Below is a step‑by‑step notebook for downloading RNA‑seq data and performing differential expression analysis to identify genes linked to nuclear import defects.

```python
import pandas as pd
import numpy as np
# Download RNA‑seq dataset from the repository
url = 'https://data_repository.org/nuclear_import_data.csv'
data = pd.read_csv(url)
print(data.head())
```

Next, apply principal component analysis (PCA) to visualize the overall variance between control and importin‑β inhibited samples.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
components = pca.fit_transform(data.select_dtypes(include=[np.number]))
import pandas as pd
pca_df = pd.DataFrame(components, columns=['PC1', 'PC2'])
print(pca_df.head())
```

Finally, create an interactive scatter plot using Plotly to visualize clustering patterns in gene expression.

```python
import plotly.express as px
fig = px.scatter(x=components[:,0], y=components[:,1], title='PCA of Nuclear Import Disruption')
fig.show()
```
