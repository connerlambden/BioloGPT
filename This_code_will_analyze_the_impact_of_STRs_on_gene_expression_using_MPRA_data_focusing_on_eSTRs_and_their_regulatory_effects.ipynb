### Step 1: Import Libraries
Import necessary libraries for data analysis.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2: Load MPRA Data
Load the MPRA dataset containing STRs and their expression levels.

```python
mpra_data = pd.read_csv('mpra_data.csv')
mpra_data.head()
```

### Step 3: Analyze STR Impact
Analyze the correlation between STR variations and gene expression levels.

```python
correlation = mpra_data.corr()
sns.heatmap(correlation, annot=True)
```
