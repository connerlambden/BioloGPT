### Data Preparation
Load datasets containing preoperative IL-6 and D-dimer levels along with postoperative outcomes.

```python
import pandas as pd

# Load datasets
df_il6 = pd.read_csv('il6_data.csv')
df_d_dimer = pd.read_csv('d_dimer_data.csv')
```

### Data Analysis
Merge datasets and calculate the integrated risk score based on IL-6 and D-dimer levels.

```python
# Merge datasets
df_merged = pd.merge(df_il6, df_d_dimer, on='patient_id')

# Calculate risk score
df_merged['risk_score'] = df_merged['IL6_level'] + df_merged['D_dimer_level']

# Analyze outcomes
outcomes = df_merged.groupby('risk_score')['outcome'].mean()
```

### Results Visualization
Visualize the relationship between the integrated risk score and postoperative outcomes.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(outcomes.index, outcomes.values)
plt.xlabel('Integrated Risk Score')
plt.ylabel('Postoperative Outcome Rate')
plt.title('Impact of Integrated Risk Score on Surgical Outcomes')
plt.show()
```
