### Analyzing Vaccination Effectiveness
This notebook will analyze the effectiveness of measles and chickenpox vaccines using historical data.

```python
import pandas as pd
# Load vaccination data
data = pd.read_csv('vaccination_data.csv')
# Analyze effectiveness
measles_effectiveness = data[data['disease'] == 'measles']['effectiveness'].mean()
chickenpox_effectiveness = data[data['disease'] == 'chickenpox']['effectiveness'].mean()
measles_effectiveness, chickenpox_effectiveness
```

### Results
The average effectiveness of the measles and chickenpox vaccines will be displayed.

```python
print(f'Measles Vaccine Effectiveness: {measles_effectiveness}%')
print(f'Chickenpox Vaccine Effectiveness: {chickenpox_effectiveness}%')
```
