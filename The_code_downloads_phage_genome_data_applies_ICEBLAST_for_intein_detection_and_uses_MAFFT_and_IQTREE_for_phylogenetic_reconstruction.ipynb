This notebook section downloads the PhagesDB dataset, pre-processes intein sequences, and performs multiple sequence alignment using MAFFT.

```python
import subprocess
subprocess.run(['mafft', '--auto', 'intein_sequences.fasta', '>', 'aligned_inteins.fasta'])
```

Next, IQTREE is used to build the phylogenetic tree from aligned sequences.

```python
subprocess.run(['iqtree', '-s', 'aligned_inteins.fasta', '-m', 'TEST', '-bb', '1000'])
```

This simple pipeline integrates intein sequence detection and phylogenetic reconstruction to validate evolutionary relationships.

```python
print('Pipeline executed: intein sequences processed and phylogeny generated')
```
