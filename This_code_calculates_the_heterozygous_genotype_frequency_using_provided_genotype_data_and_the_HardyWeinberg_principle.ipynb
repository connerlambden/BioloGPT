### Step 1: Define the genotype frequencies
- AA frequency = 0.25
- aa frequency = 0.35

### Step 2: Calculate allele frequencies
- p = sqrt(AA frequency)
- q = sqrt(aa frequency)

### Step 3: Calculate heterozygous frequency
- Heterozygous frequency (Aa) = 2 * p * q

```python
# Given frequencies
AA_freq = 0.25
aa_freq = 0.35

# Calculate allele frequencies
import math
p = math.sqrt(AA_freq)
q = math.sqrt(aa_freq)

# Calculate heterozygous frequency
heterozygous_freq = 2 * p * q
heterozygous_freq
```

### Step 4: Output the result
- The calculated frequency of the heterozygous genotype (Aa) is approximately: 

```python
print(f'Heterozygous frequency (Aa): {heterozygous_freq:.4f}')
```
