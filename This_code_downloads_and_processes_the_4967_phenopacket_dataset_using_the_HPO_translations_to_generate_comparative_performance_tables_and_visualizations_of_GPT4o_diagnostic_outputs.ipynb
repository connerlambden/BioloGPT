The notebook begins by loading the phenopacket dataset, applying HPO translations, and mapping diagnoses using the Mondo disease ontology to generate performance metrics.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Download dataset from provided link
url = 'https://zenodo.org/records/14907503'
df = pd.read_csv(url)
# Process the data for performance comparisons
table = df.groupby('language')[['top1_accuracy', 'top3_accuracy']].mean()
print(table)
plt.figure(figsize=(8,4))
table.plot(kind='bar')
plt.title('Mean Diagnostic Performance by Language')
plt.xlabel('Language')
plt.ylabel('Accuracy (%)')
plt.show()
```

The notebook further analyzes statistical differences using non-parametric tests and outputs an interactive graph for deeper exploration.

```python
from scipy.stats import kruskal
# Assuming performance_data is a dict of lists for each language
data = table['top1_accuracy'].tolist()
stat, p = kruskal(*data)
print('Kruskal-Wallis H-test: Statistic =', stat, ', p-value =', p)
```
