### Phylogenetic Tree Construction Using Biopython
This notebook will guide you through the process of constructing a phylogenetic tree using genetic data.

```python
from Bio import Phylo
from Bio.Align import MultipleSeqAlignment
from Bio import SeqIO

# Load sequence data
sequences = SeqIO.parse('path_to_sequences.fasta', 'fasta')

# Create a multiple sequence alignment
alignment = MultipleSeqAlignment(sequences)

# Construct a phylogenetic tree using the Neighbor-Joining method
from Bio.Phylo.TreeConstruction import DistanceCalculator, DistanceTreeConstructor
calculator = DistanceCalculator('identity')
dist_matrix = calculator.get_distance(alignment)
constructor = DistanceTreeConstructor()
tree = constructor.nj(dist_matrix)

# Visualize the tree
Phylo.draw(tree)
```
