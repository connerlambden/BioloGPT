This notebook will analyze HMGB1 expression levels in response to rotenone treatment using RNA-seq data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
hmgb1_data = pd.read_csv('hmgb1_expression_data.csv')

# Visualize HMGB1 expression levels
sns.boxplot(x='treatment', y='expression', data=hmgb1_data)
plt.title('HMGB1 Expression in Rotenone-Treated Cells')
plt.show()
```

The analysis will provide insights into how rotenone affects HMGB1 expression and its potential implications for DNA repair.
