### Data Analysis of RPS6KB1 and PDPK1 Expression Levels

```python
import pandas as pd

# Load cancer expression dataset
dataset = pd.read_csv('cancer_expression_data.csv')

# Analyze expression levels
rps6kb1_expression = dataset['RPS6KB1']
pdpk1_expression = dataset['PDPK1']

# Correlation analysis
correlation = rps6kb1_expression.corr(pdpk1_expression)
print('Correlation between RPS6KB1 and PDPK1:', correlation)
```
