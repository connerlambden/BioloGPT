This notebook will analyze the stability of urea based on pH and temperature data.

```python
import pandas as pd

# Load stability data for urea
urea_data = pd.read_csv('urea_stability_data.csv')
urea_data.head()
```

The data includes pH levels, temperature, and corresponding stability measurements.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(urea_data['pH'], urea_data['stability'], marker='o')
plt.title('Urea Stability vs pH')
plt.xlabel('pH')
plt.ylabel('Stability')
plt.grid()
plt.show()
```

This graph visualizes the relationship between pH and urea stability.
