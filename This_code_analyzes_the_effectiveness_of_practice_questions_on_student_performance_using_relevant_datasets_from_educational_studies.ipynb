### Analyzing the Impact of Practice Questions on Student Performance
This notebook will utilize datasets from various studies to evaluate how practice questions influence exam scores.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data representing exam scores
data = {'Year': [2012, 2013, 2014, 2015], 'Exam_1': [88.7, 89.1, 95.3, 104.9], 'Exam_2': [98.3, 94.0, 105.7, 83.6]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Year'], df['Exam_1'], label='Exam 1')
plt.bar(df['Year'], df['Exam_2'], label='Exam 2', bottom=df['Exam_1'])
plt.title('Exam Scores Over Years')
plt.xlabel('Year')
plt.ylabel('Scores')
plt.legend()
plt.show()
```
