This notebook will analyze RNA-seq data from PKR mutant cell lines to identify changes in gene expression.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load the dataset
data = pd.read_csv('PKR_mutant_expression_data.csv')

# Perform differential expression analysis
results = stats.ttest_ind(data['mutant'], data['wildtype'])

# Output results
results_df = pd.DataFrame({'Gene': data['Gene'], 'p-value': results.pvalue})
results_df.to_csv('differential_expression_results.csv', index=False)
```
