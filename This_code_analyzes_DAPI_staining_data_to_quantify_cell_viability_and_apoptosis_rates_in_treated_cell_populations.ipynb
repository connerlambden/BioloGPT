### DAPI Staining Data Analysis
This notebook will guide you through the analysis of DAPI staining data to quantify cell viability and apoptosis.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load DAPI staining data
data = pd.read_csv('dapi_staining_data.csv')

# Analyze cell viability
viability = data['viable_cells'] / data['total_cells'] * 100
plt.hist(viability, bins=20)
plt.title('Cell Viability Distribution')
plt.xlabel('Viability (%)')
plt.ylabel('Frequency')
plt.show()
```

### Results Interpretation
The histogram above shows the distribution of cell viability percentages across the samples.
