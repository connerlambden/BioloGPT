### Gene Expression Analysis for Hematopoietic Differentiation

```python
import pandas as pd

data = pd.read_csv('gene_expression_data.csv')
# Analyze expression levels of CD34 and CD45 markers
cd34_expression = data[data['gene'] == 'CD34']
cd45_expression = data[data['gene'] == 'CD45']

# Visualize the expression levels
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(cd34_expression['time'], cd34_expression['expression'], label='CD34')
plt.plot(cd45_expression['time'], cd45_expression['expression'], label='CD45')
plt.xlabel('Time (days)')
plt.ylabel('Expression Level')
plt.title('Gene Expression Over Time')
plt.legend()
plt.show()
```

### Discussion
This analysis helps identify the optimal time points for harvesting cells based on marker expression.
