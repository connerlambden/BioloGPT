This notebook will analyze DNA sequences for base composition and identify any modifications present.

```python
import pandas as pd

def analyze_dna_sequence(sequence):
    base_counts = {base: sequence.count(base) for base in 'ATCG'}
    return base_counts

# Example usage
sequence = 'ATCGATCGATCG'
result = analyze_dna_sequence(sequence)
print(result) # Output: {'A': 3, 'T': 3, 'C': 3, 'G': 3}
```
