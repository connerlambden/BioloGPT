This notebook analyzes RNA-seq data from iRMD patients to identify key gene expression changes.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load transcriptomic data
data = pd.read_csv('iRMD_transcriptomic_data.csv')

# Perform differential expression analysis
results = stats.ttest_ind(data['iRMD'], data['control'])

# Output results
results_df = pd.DataFrame({'Gene': data['Gene'], 'p-value': results.pvalue})
results_df.to_csv('differential_expression_results.csv', index=False)
```
