This notebook will analyze RNA-seq data from Mtr4 knockout and control samples to identify differentially expressed genes associated with meiotic initiation.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load RNA-seq data
control_data = pd.read_csv('control_rna_seq.csv')
knockout_data = pd.read_csv('knockout_rna_seq.csv')

# Perform differential expression analysis
merged_data = pd.merge(control_data, knockout_data, on='gene_id')
merged_data['log2_fold_change'] = np.log2(merged_data['knockout_expression'] / merged_data['control_expression'])

# Identify significantly differentially expressed genes
significant_genes = merged_data[merged_data['p_value'] < 0.05]
```

The analysis identifies genes that are significantly upregulated or downregulated in Mtr4 knockout samples compared to controls, providing insights into MTR4's role in meiotic initiation.

```python
# Output significant genes
significant_genes.to_csv('significant_genes.csv', index=False)
```
