### Analysis of Hypothetical Genes in Phage S4

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load genomic data of phage S4
phage_data = pd.read_csv('phage_s4_genome.csv')

# Analyze the distribution of hypothetical genes
hypothetical_genes = phage_data[phage_data['type'] == 'hypothetical']

# Plot the distribution
plt.figure(figsize=(10,6))
plt.hist(hypothetical_genes['length'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Hypothetical Gene Lengths in Phage S4')
plt.xlabel('Gene Length (bp)')
plt.ylabel('Frequency')
plt.show()
```
