Download and pre-process patient RNA-seq datasets from the provided GitHub repository. Filter for RTN3, CRTH2, and collagen gene expression levels.

```python
import pandas as pd
# Load dataset from GitHub URL
url = 'https://raw.githubusercontent.com/alfredsguo/ipf_wes_analysis/main/rnaseq_data.csv'
data = pd.read_csv(url)
# Filter relevant columns
filtered_data = data[['RTN3', 'CRTH2', 'COL1A1', 'COL1A2']]
filtered_data.corr()
print(filtered_data.corr())
```

Generate correlation graphs using Plotly to visualize relationships between RTN3 and fibrosis markers.

```python
import plotly.express as px
fig = px.scatter(filtered_data, x='RTN3', y='COL1A1', title='RTN3 vs COL1A1 Expression')
fig.show()
```
