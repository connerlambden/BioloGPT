### Step 1: Import Libraries
Import necessary libraries for data analysis and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for protein abundance changes
data = {'Protein': ['SOD 1', 'GR', 'Pathogenesis-related protein 1', 'Bowman-Birk proteinase inhibitor'],
        'Abundance Change': [1.5, 1.3, 0.5, 0.5]}

# Create DataFrame
df = pd.DataFrame(data)

# Step 2: Visualize Data
plt.bar(df['Protein'], df['Abundance Change'], color='blue')
plt.title('Protein Abundance Changes Under Salt Stress and MMW Irradiation')
plt.xlabel('Proteins')
plt.ylabel('Relative Abundance')
plt.show()
```
