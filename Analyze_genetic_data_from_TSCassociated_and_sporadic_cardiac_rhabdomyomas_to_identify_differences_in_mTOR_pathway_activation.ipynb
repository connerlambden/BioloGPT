This notebook will analyze genetic mutations in cardiac rhabdomyomas.

```python
# Load necessary libraries
import pandas as pd
import numpy as np

# Load datasets
# Example: TSC-associated mutations
TSC_data = pd.read_csv('TSC_mutations.csv')
# Example: Sporadic mutations
sporadic_data = pd.read_csv('sporadic_mutations.csv')
```

### Analyze mutation frequencies and compare mTOR pathway involvement.

```python
# Analyze mutation frequencies
TSC_mutation_counts = TSC_data['mutation'].value_counts()
sporadic_mutation_counts = sporadic_data['mutation'].value_counts()
```

### Visualize the differences in mutation frequencies.

```python
import matplotlib.pyplot as plt

# Plotting mutation frequencies
plt.figure(figsize=(10,5))
plt.bar(TSC_mutation_counts.index, TSC_mutation_counts.values, alpha=0.5, label='TSC-associated')
plt.bar(sporadic_mutation_counts.index, sporadic_mutation_counts.values, alpha=0.5, label='Sporadic')
plt.xlabel('Mutations')
plt.ylabel('Frequency')
plt.title('Mutation Frequencies in Cardiac Rhabdomyomas')
plt.legend()
plt.show()
```
