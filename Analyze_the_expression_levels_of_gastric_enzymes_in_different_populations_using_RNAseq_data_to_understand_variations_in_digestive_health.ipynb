This analysis will utilize RNA-seq data to compare the expression levels of pepsin and gastric lipase across different dietary groups.

```python
import pandas as pd
# Load RNA-seq data
rna_seq_data = pd.read_csv('gastric_enzyme_expression.csv')
# Analyze expression levels
expression_summary = rna_seq_data.groupby('diet').mean()
```

The results will provide insights into how diet influences the expression of gastric enzymes.

```python
expression_summary.to_csv('expression_summary.csv')
```
