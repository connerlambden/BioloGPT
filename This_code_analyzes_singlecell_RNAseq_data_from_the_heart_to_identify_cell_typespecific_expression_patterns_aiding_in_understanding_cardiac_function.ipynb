This notebook explores single-cell RNA-seq datasets to identify distinct cell types in the heart.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('heart_single_cell_data.csv')

# Analyze expression patterns
sns.boxplot(x='cell_type', y='gene_expression', data=data)
plt.title('Gene Expression by Cell Type in Heart')
plt.show()
```

The analysis reveals distinct expression patterns across different heart cell types, providing insights into cardiac function.
