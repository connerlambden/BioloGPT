Análisis de datos de expresión génica en procariotas.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar datos de expresión génica
exp_data = pd.read_csv('data.csv')
sns.boxplot(data=exp_data) 
plt.title('Expresión génica en procariotas')
plt.show()
```

Visualización de la expresión génica en diferentes condiciones.

```python
# Visualización de datos
sns.heatmap(exp_data.corr(), annot=True)
plt.title('Correlación de expresión génica')
plt.show()
```
