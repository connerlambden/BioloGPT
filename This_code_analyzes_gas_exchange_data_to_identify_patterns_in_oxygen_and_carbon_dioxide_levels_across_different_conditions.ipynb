### Gas Exchange Data Analysis
This notebook analyzes the gas exchange data to identify trends and correlations.

```python
import pandas as pd

data = {'Gas': ['Oxygen', 'Carbon Dioxide'], 'Partial Pressure (mmHg)': [100, 40]}
df = pd.DataFrame(data)
df.describe()
```

### Summary Statistics
The summary statistics provide insights into the gas exchange efficiency.

```python
df.plot(kind='bar', x='Gas', y='Partial Pressure (mmHg)', title='Gas Exchange Partial Pressures')
```
