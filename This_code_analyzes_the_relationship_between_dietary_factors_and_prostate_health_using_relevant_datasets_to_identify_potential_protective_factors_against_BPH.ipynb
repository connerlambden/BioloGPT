### Data Analysis on Dietary Factors and Prostate Health

```python
import pandas as pd

diet_data = pd.read_csv('dietary_factors.csv')
prostate_data = pd.read_csv('prostate_health.csv')

# Merge datasets on common identifiers
merged_data = pd.merge(diet_data, prostate_data, on='patient_id')

# Analyze correlation between dietary fiber intake and prostate volume
correlation = merged_data['fiber_intake'].corr(merged_data['prostate_volume'])
print('Correlation between fiber intake and prostate volume:', correlation)
```
