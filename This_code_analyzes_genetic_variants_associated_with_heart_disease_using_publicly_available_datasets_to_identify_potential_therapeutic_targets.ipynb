This notebook will analyze genetic variants linked to heart disease using GWAS data.

```python
import pandas as pd
# Load datasets
variants = pd.read_csv('path_to_variants_data.csv')
# Analyze associations
associations = variants.groupby('gene').agg({'risk': 'mean'})
```

The analysis will provide insights into which genetic variants are most strongly associated with heart disease.

```python
# Display results
associations.sort_values(by='risk', ascending=False).head(10)
```
