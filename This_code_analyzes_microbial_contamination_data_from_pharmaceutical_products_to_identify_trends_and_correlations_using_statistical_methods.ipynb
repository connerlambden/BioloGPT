This notebook will analyze microbial contamination data, focusing on identifying trends and correlations in contamination rates.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load contamination data
data = pd.read_csv('contamination_data.csv')

# Analyze trends in contamination rates
data.groupby('product_type')['contamination_rate'].mean().plot(kind='bar')
plt.title('Average Contamination Rates by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Average Contamination Rate')
plt.show()
```

The analysis provides insights into which product types are most affected by microbial contamination, guiding future sampling strategies.
