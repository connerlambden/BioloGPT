### Analysis of OGTT Adherence Data
This notebook will analyze the adherence rates to OGTT screening among TDT patients based on data collected from various centers.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'City': ['Antalya', 'Ankara', 'Izmir', 'Adana', 'Muscat', 'Colombo'],
    'Adherence_Percentage': [70, 90, 90, 10, 50, 10]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['City'], df['Adherence_Percentage'], color='skyblue')
plt.title('Adherence Rates to OGTT Screening by City')
plt.xlabel('City')
plt.ylabel('Adherence Percentage (%)')
plt.show()
```

### Discussion
This analysis provides insights into the adherence rates across different cities, highlighting areas that may require targeted interventions.
