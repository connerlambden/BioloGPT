This notebook reads the experimental dataset provided from the paper and generates visualizations to compare weight loss and thickness reduction across specimens.

```python
import plotly.graph_objects as go
import pandas as pd

# Create a dataframe from the extracted data
data = {
    'Specimen': ['CTS/Coll', 'CTS/Coll + 5%PS', 'CTS/Coll + 10%PS', 'CTS/Coll/5TA + 5%PS', 'CTS/Coll/5TA + 10%PS', 'CTS/Coll/20TA + 5%PS', 'CTS/Coll/20TA + 10%PS'],
    'WeightLoss': [38.69, 25.01, 21.34, 34.52, 19.08, 5.26, 5.88],
    'ThicknessBefore': [0.0867, 0.0730, 0.0607, 0.0827, 0.0643, 0.0750, 0.0690],
    'ThicknessAfter': [0.0563, 0.0513, 0.0473, 0.0557, 0.0527, 0.0687, 0.0640]
}
df = pd.DataFrame(data)
df['ThicknessReduction'] = df['ThicknessBefore'] - df['ThicknessAfter']

# Plot weight loss and thickness reduction
fig = go.Figure(data=[
    go.Bar(name='Weight Loss (%)', x=df['Specimen'], y=df['WeightLoss'], marker_color='#6A0C76'),
    go.Bar(name='Thickness Reduction (mm)', x=df['Specimen'], y=df['ThicknessReduction'], marker_color='#a64dff')
])
fig.update_layout(barmode='group', title='Experimental Data Comparison', xaxis_title='Specimen', yaxis_title='Value')
fig.show()
```

This chart provides a comparative visualization of weight loss and film thickness reduction, helping to assess the material enhancement effects due to TA and PS integration.
