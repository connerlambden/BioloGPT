This notebook analyzes force and work data from various fruits to understand the mechanical performance of frugivorous fishes.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
fruits = {'Fruit': ['Macadamia nuts', 'Kiwis', 'Cherries'], 'Load at Puncture (N)': [92.27, 0, 0], 'Work to Puncture (Nm)': [4.17, 0, 0]}
df = pd.DataFrame(fruits)

# Plotting
plt.bar(df['Fruit'], df['Load at Puncture (N)'], color='blue', label='Load at Puncture (N)')
plt.bar(df['Fruit'], df['Work to Puncture (Nm)'], color='orange', label='Work to Puncture (Nm)', alpha=0.7)
plt.title('Force and Work Required to Puncture Fruits')
plt.xlabel('Fruit Type')
plt.ylabel('Force (N) / Work (Nm)')
plt.legend()
plt.show()
```

This analysis provides insights into how different fruits require varying amounts of force and work to be processed by frugivorous fishes.
