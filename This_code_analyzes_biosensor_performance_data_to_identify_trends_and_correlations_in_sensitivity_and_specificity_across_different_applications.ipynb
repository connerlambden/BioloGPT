### Biosensor Performance Analysis
This notebook analyzes the performance of various commercial biosensors based on sensitivity and specificity.

```python
import pandas as pd

# Sample data for biosensor performance
biosensor_data = {
    'Biosensor': ['FreeStyle Life', 'QuickVue', 'HATOX-2000'],
    'Sensitivity (%)': [95, 94, 98],
    'Specificity (%)': [90, 85, 95]
}
df = pd.DataFrame(biosensor_data)
df
```

### Summary of Biosensor Performance
The table above summarizes the sensitivity and specificity of selected commercial biosensors.

```python
# Plotting the performance
import matplotlib.pyplot as plt

plt.bar(df['Biosensor'], df['Sensitivity (%)'], color='blue', label='Sensitivity')
plt.bar(df['Biosensor'], df['Specificity (%)'], color='orange', label='Specificity', alpha=0.7)
plt.ylabel('Percentage (%)')
plt.title('Biosensor Performance Comparison')
plt.legend()
plt.show()
```
