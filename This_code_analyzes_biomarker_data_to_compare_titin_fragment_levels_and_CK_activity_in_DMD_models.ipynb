### Analysis of Biomarker Sensitivity in DMD Models
This notebook will analyze the differences in urine titin fragment levels and plasma CK activity across various DMD mouse models.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for analysis
data = {
    'Mouse Group': ['DMD-null', 'DYS-HAC1; DMD-null', 'WT'],
    'Urine Titin Fragment Levels': [1.5, 1.2, 0.5],
    'Plasma CK Activity': [200, 150, 100]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['Mouse Group'], df['Urine Titin Fragment Levels'], label='Urine Titin Fragment Levels')
plt.bar(df['Mouse Group'], df['Plasma CK Activity'], label='Plasma CK Activity', alpha=0.7)
plt.title('Comparison of Biomarkers in DMD Mouse Models')
plt.xlabel('Mouse Groups')
plt.ylabel('Biomarker Levels')
plt.legend()
plt.show()
```
