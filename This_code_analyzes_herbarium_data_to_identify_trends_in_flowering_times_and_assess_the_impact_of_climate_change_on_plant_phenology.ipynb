### Analyzing Herbarium Data for Phenological Trends
This notebook will analyze flowering time data from herbarium specimens to identify trends and correlations with climate variables.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load herbarium data
herbarium_data = pd.read_csv('herbarium_flowering_data.csv')

# Analyze trends in flowering times
def analyze_flowering_times(data):
    plt.figure(figsize=(10, 5))
    plt.plot(data['Year'], data['Flowering Peak'], marker='o')
    plt.title('Trends in Flowering Times')
    plt.xlabel('Year')
    plt.ylabel('Flowering Peak')
    plt.grid()
    plt.show()

analyze_flowering_times(herbarium_data)
```
