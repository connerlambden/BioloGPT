```python
# Código para buscar TFMs en un repositorio académico
import requests

# Función para buscar TFMs
def buscar_tfm(titulo):
    url = f'https://api.repositorio.com/search?query={titulo}'
    response = requests.get(url)
    return response.json()
```

Este código permite realizar búsquedas en un repositorio académico para encontrar TFMs relevantes.

```python
# Ejemplo de búsqueda
resultados = buscar_tfm('neurociencia')
print(resultados)
```
