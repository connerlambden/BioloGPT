### Exoplanet Data Analysis
This notebook will analyze the distribution of exoplanet types and their orbital characteristics using data from recent surveys.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load exoplanet data
exoplanet_data = pd.read_csv('exoplanet_data.csv')

# Analyze distribution of exoplanet types
exoplanet_types = exoplanet_data['type'].value_counts()

# Plot the distribution
plt.figure(figsize=(10,6))
exoplanet_types.plot(kind='bar')
plt.title('Distribution of Exoplanet Types')
plt.xlabel('Exoplanet Type')
plt.ylabel('Count')
plt.show()
```

This analysis provides insights into the prevalence of different exoplanet types, which can inform models of planetary formation.
