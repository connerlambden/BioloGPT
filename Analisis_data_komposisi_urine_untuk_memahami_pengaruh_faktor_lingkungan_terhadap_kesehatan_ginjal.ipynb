Analisis data komposisi urine berdasarkan suhu dan kelembapan.

```python
import pandas as pd
# Load data
urine_data = pd.read_csv('urine_composition.csv')
# Analyze correlations
correlation = urine_data.corr()
```

Analisis hasil untuk menentukan hubungan antara suhu, kelembapan, dan komposisi urine.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(correlation, annot=True)
plt.title('Korelasi antara Suhu, Kelembapan, dan Komposisi Urine')
plt.show()
```
