This section details the annotation of transcriptomic profiles comparing microglia and astrocytes under high-salt conditions.

```python
import scanpy as sc
adata = sc.read('high_salt_microglia.h5ad')
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.tl.pca(adata, svd_solver='arpack')
sc.pl.pca_variance_ratio(adata, log=True)
```

The above code loads the dataset, filters and normalizes the data, then conducts PCA to explore variability associated with high-salt induced changes.

```python
sc.tl.rank_genes_groups(adata, 'condition', method='t-test')
sc.pl.rank_genes_groups(adata, n_genes=20, sharey=False)
```

This segment identifies differentially expressed genes between conditions, a key step in linking transcriptomic changes to glial remodeling.
