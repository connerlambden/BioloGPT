### Biocompatibility Analysis of Scaffolds
This analysis will evaluate the MTT assay results for different scaffold compositions.

```python
import pandas as pd
import scipy.stats as stats

# Sample data for MTT assay results
data = {'Scaffold': ['GO/SCAFFG', 'GO/SCAFFGA', 'GO/SCAFFG3A'], 'Cell Viability (%)': [85, 90, 80]}
df = pd.DataFrame(data)

# Perform ANOVA test
f_statistic, p_value = stats.f_oneway(df['Cell Viability (%)'])

print(f'F-statistic: {f_statistic}, P-value: {p_value}')
```

### Results Interpretation
A p-value < 0.05 indicates significant differences in cell viability among the scaffold types.
