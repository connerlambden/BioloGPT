Analyze arpin gene expression data from endothelial cells.

```python
import pandas as pd
# Load gene expression data
expression_data = pd.read_csv('arpin_expression.csv')
```

Identify significant changes in expression under inflammatory conditions.

```python
significant_changes = expression_data[expression_data['p_value'] < 0.05]
```

Visualize the expression changes.

```python
import matplotlib.pyplot as plt
plt.bar(significant_changes['condition'], significant_changes['expression_level'])
plt.title('Arpin Expression Changes')
plt.show()
```
