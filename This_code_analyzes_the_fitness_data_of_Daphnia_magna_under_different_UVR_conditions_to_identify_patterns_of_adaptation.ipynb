### Data Analysis of Daphnia magna Fitness Traits Under UVR Variability

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = 'https://doi.org/10.5061/dryad.m63xsj4cw'
data = pd.read_csv(url)

# Analyze survival rates under different UVR conditions
survival_rates = data.groupby('UVR_condition')['survival'].mean()

# Plotting the results
plt.figure(figsize=(10, 6))
plt.bar(survival_rates.index, survival_rates.values)
plt.title('Survival Rates of Daphnia magna Under Different UVR Conditions')
plt.xlabel('UVR Condition')
plt.ylabel('Average Survival Rate')
plt.show()
```
