In this notebook, we load sequence metadata and ASV count tables from the study to compare single-step and nested PCR results. We employ pandas for data handling and generate summary statistics.

```python
import pandas as pd
import numpy as np

# Load dataset (replace 'data.csv' with actual dataset path)
df = pd.read_csv('data.csv')

# Summary statistics of ASV counts
summary_stats = df.describe()
print(summary_stats)

# Compare amplification efficiency between nested and single-step PCR
nested = df[df['Method'] == 'Nested']
single_step = df[df['Method'] == 'Single']

efficiency_diff = nested['Amplicon_Count'].mean() - single_step['Amplicon_Count'].mean()
print('Difference in amplification efficiency:', efficiency_diff)
```

The above code computes basic descriptive statistics and compares mean amplification counts between methods to assess sensitivity improvements.

```python
import matplotlib.pyplot as plt

# Plotting amplification efficiency for Nested vs Single-step
fig, ax = plt.subplots()
methods = ['Nested', 'Single']
means = [nested['Amplicon_Count'].mean(), single_step['Amplicon_Count'].mean()]

ax.bar(methods, means, color=['#6A0C76', '#CCCCCC'])
ax.set_title('Amplification Efficiency Comparison')
ax.set_ylabel('Mean Amplicon Count')

plt.show()
```

This visualization provides a clear comparison of the two methods, reinforcing the sensitivity improvements of the nested PCR approach.
