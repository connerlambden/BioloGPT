This notebook will analyze gene expression profiles of T cells from thymic and extrathymic sources.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('t_cell_expression_data.csv')
sns.boxplot(x='source', y='gene_expression', data=data)
plt.title('T Cell Gene Expression by Source')
plt.show()
```

The boxplot above illustrates the differences in gene expression between thymic and extrathymic T cells.
