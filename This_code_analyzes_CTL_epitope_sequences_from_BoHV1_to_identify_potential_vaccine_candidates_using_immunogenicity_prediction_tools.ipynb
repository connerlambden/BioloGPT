This notebook will utilize datasets of known CTL epitopes to predict the immunogenicity of potential vaccine candidates against BoHV-1.

```python
import pandas as pd
# Load CTL epitope data
epitopes = pd.read_csv('ctl_epitopes.csv')
# Analyze immunogenicity scores
immunogenicity_scores = epitopes['VaxiJen_score']
# Filter for high-scoring epitopes
high_scoring = epitopes[immunogenicity_scores > 1.0]
```

The analysis will provide a list of high-scoring CTL epitopes that can be further evaluated for vaccine development.

```python
high_scoring.to_csv('high_scoring_epitopes.csv', index=False)
```
