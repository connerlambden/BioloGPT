This notebook details the analysis of heavy metal reduction and differential gene expression in rice treated with [Glu][H2PO4], using the provided experimental datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Parameter': ['Cd X24', 'As X24', 'Cd Z35', 'As Z35'],
    'Before': [0.41, 0.82, 0.48, 0.93],
    'After': [0.17, 0.51, 0.2, 0.52]
})

data['Reduction(%)'] = 100 * (data['Before'] - data['After']) / data['Before']

data.plot(kind='bar', x='Parameter', y='Reduction(%)', color='#6A0C76', title='Heavy Metal Reduction in Rice')
plt.ylabel('Reduction (%)')
plt.show()
```

The above visualization confirms the reported reductions in Cd and As content post treatment.
