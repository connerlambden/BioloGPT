## Correlation Analysis between MIC and CC50

This notebook analyzes the relationship between the minimum inhibitory concentration (MIC) and cytotoxic concentration (CC50) of the synthesized compounds to identify those with the best therapeutic index.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Creating DataFrame from extracted data
data = {
    'Compound': ['MNS2', 'MNS3', 'MNS4', 'MNS5', 'MNS6'],
    'MIC_with_Gentamicin': [100, 90, 60, 60, 100],
    'CC50_HEK273T': [25, 190, 87, 250, 260]
}
df = pd.DataFrame(data)

# Display the DataFrame
df
```

### Scatter Plot of MIC vs CC50

This plot illustrates the relationship between the MIC values and CC50 for each compound, highlighting their therapeutic indices.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='MIC_with_Gentamicin', y='CC50_HEK273T', hue='Compound', s=100)
plt.title('Correlation between MIC and CC50')
plt.xlabel('MIC with Gentamicin (µM)')
plt.ylabel('CC50 HEK273T (µM)')
plt.legend(title='Compound')
plt.show()
```

### Interpretation

Compounds with lower MIC values and higher CC50 values, such as MNS3 and MNS4, demonstrate a favorable therapeutic index, indicating their potential as effective and safe antibiotic potentiators.
