### Kinetic Data Analysis for CRISPR-Cas9
This notebook analyzes kinetic data to identify potential mutations in the HNH domain that could enhance specificity.

```python
import pandas as pd
import numpy as np

# Load kinetic data
kinetic_data = pd.read_csv('kinetic_data.csv')

# Analyze the effects of mutations on specificity
results = kinetic_data.groupby('mutation').agg({'specificity':'mean', 'efficiency':'mean'})
results.sort_values(by='specificity', ascending=False)
```
