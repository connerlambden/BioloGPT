This notebook will analyze genetic sequences from wild and cultivated <i>Lobularia maritima</i> to identify key mutations.

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load genetic data
wild_data = pd.read_csv('wild_lobularia_genetics.csv')
cultivated_data = pd.read_csv('cultivated_lobularia_genetics.csv')

# Analyze mutations
mutations = wild_data.merge(cultivated_data, on='gene_id', suffixes=('_wild', '_cultivated'))

# Identify significant mutations
significant_mutations = mutations[mutations['expression_cultivated'] > mutations['expression_wild']]

# Output results
significant_mutations.to_csv('significant_mutations.csv')
```

This analysis will help identify genetic changes that have occurred during the domestication of <i>Lobularia maritima</i>.

```python
# Display significant mutations
print(significant_mutations)
```
