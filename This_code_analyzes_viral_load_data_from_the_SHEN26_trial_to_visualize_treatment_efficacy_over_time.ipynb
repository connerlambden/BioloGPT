### Viral Load Analysis of SHEN26 Clinical Trial

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from the trial
viral_load_data = {
    'Group': ['Placebo', '200 mg', '400 mg'],
    'Day 3': [-1.93, -2.08, -2.99],
    'Day 5': [-3.12, -3.22, -4.33]
}

# Create DataFrame
df = pd.DataFrame(viral_load_data)

# Plotting
plt.figure(figsize=(10, 6))
for day in ['Day 3', 'Day 5']:
    plt.plot(df['Group'], df[day], marker='o', label=day)

plt.title('Viral Load Reduction by Treatment Group')
plt.xlabel('Treatment Group')
plt.ylabel('Viral Load (log10 copies/mL)')
plt.legend()
plt.grid()
plt.show()
```
