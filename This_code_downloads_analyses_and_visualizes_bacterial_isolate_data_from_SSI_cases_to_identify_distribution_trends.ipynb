```python
import plotly.graph_objs as go
import plotly.offline as pyo

bacteria = ['E coli', 'S aureus', 'P aeruginosa', 'K pneumoniae', 'M morganii', 'E cloacae', 'A baumannii']
counts = [11, 7, 4, 1, 2, 1, 1]

trace = go.Bar(x=bacteria, y=counts, marker=dict(color='#6A0C76'))
layout = go.Layout(title='Bacterial Isolates from SSI Cases', xaxis=dict(title='Bacterial Species'), yaxis=dict(title='Count'))
fig = go.Figure(data=[trace], layout=layout)
pyo.plot(fig, filename='ssi_bacteria_distribution.html')
```
