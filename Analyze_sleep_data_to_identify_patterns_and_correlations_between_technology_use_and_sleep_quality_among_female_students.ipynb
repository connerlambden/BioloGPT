This notebook will analyze sleep data collected from female university students in the UAE, focusing on technology use and sleep quality.

```python
import pandas as pd

# Load sleep data
sleep_data = pd.read_csv('sleep_data.csv')

# Analyze correlations
correlation_matrix = sleep_data.corr()
```

The correlation matrix will help identify significant relationships between sleep quality, technology use, and other variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Sleep Data')
plt.show()
```
