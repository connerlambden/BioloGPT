### Impedance Data Analysis for Caco-2 Cells

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample impedance data
impedance_data = {
    'Day': ['Day 0', 'Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5', 'Day 6', 'Day 7', 'Day 8', 'Day 9', 'Day 10', 'Day 11', 'Day 12', 'Day 13', 'Day 14'],
    'Normalized_Impedance_Change': [0, 100, 214, 300, 350, 400, 450, 453, 400, 380, 360, 340, 320, 300, 280]
}

# Create DataFrame
df = pd.DataFrame(impedance_data)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(df['Day'], df['Normalized_Impedance_Change'], marker='o')
plt.title('Normalized Change in Impedance of Caco-2 Cells Over Time')
plt.xlabel('Days in Culture')
plt.ylabel('Normalized Impedance Change (%)')
plt.grid()
plt.show()
```
