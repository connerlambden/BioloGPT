This notebook analyzes the dataset extracted from the paper to correlate flow rate with fiber diameter and porosity outcomes.

```python
import pandas as pd
import plotly.express as px

data = pd.DataFrame({
    'Flow_Rate': [10,20,30,40,50,60,80,100],
    'Measured_Diameter': [73,123,164,193,252,277,366,430],
    'Porosity': [89.22,79.51,69.4,60.24,49.14,39.87,20.26,0]
})

fig = px.line(data, x='Flow_Rate', y=['Measured_Diameter', 'Porosity'],
              title='Fiber Diameter and Porosity vs Flow Rate',
              labels={'value':'Value', 'Flow_Rate':'Flow Rate Percentage'})
fig.show()
```

The above code downloads the dataset, constructs a DataFrame, and creates a line plot to visualize the relationship between flow rates, fiber diameters, and porosity.
