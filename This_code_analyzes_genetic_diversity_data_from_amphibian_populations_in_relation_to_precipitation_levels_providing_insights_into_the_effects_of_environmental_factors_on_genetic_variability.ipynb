### Analysis of Genetic Diversity in Relation to Precipitation
This notebook will analyze genetic diversity data from amphibian populations, correlating it with precipitation levels during the driest quarter.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load genetic diversity and precipitation data
data = pd.read_csv('amphibian_genetic_diversity.csv')

# Analyze correlation between precipitation and genetic diversity
correlation = data['precipitation'].corr(data['genetic_diversity'])

# Visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='precipitation', y='genetic_diversity', data=data)
plt.title('Genetic Diversity vs. Precipitation')
plt.xlabel('Precipitation (mm)')
plt.ylabel('Genetic Diversity (Heterozygosity)')
plt.show()
```
