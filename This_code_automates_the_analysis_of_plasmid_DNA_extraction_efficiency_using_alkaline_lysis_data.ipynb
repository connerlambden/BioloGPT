### Analysis of Plasmid DNA Extraction Efficiency
This notebook analyzes the yield of plasmid DNA from various bacterial cultures using the alkaline lysis method.

```python
import pandas as pd

data = {'Culture Volume (mL)': [1.5, 500], 'Plasmid Yield (µg)': [5, 4000]}

df = pd.DataFrame(data)
print(df) # Display the yield data.
```

### Yield Analysis
The above data shows the yield of plasmid DNA from different culture volumes.

```python
import matplotlib.pyplot as plt

plt.bar(df['Culture Volume (mL)'], df['Plasmid Yield (µg)'], color='purple')
plt.xlabel('Culture Volume (mL)')
plt.ylabel('Plasmid Yield (µg)')
plt.title('Plasmid DNA Yield from Alkaline Lysis')
plt.show()
```
