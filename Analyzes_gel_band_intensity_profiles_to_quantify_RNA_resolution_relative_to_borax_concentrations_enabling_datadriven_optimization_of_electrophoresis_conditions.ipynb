This notebook downloads experimental gel data and processes band intensity values to correlate resolution with borax concentration using Plotly for visualization.

```python
import pandas as pd
import numpy as np
import plotly.express as px

# Load gel intensity data; expected columns: borax_concentration, resolution_score
data = pd.read_csv('data.csv')

# Create scatter plot with a lowess trendline showing how resolution varies with borax concentration
fig = px.scatter(data, x='borax_concentration', y='resolution_score', trendline='lowess', title='RNA Resolution vs Borax Concentration')
fig.update_layout(title_font_color='#6A0C76')
fig.show()

# Calculate correlation coefficient
corr = np.corrcoef(data['borax_concentration'], data['resolution_score'])[0,1]
print('Correlation:', corr)
```

This code generates a dynamic plot that visualizes the relationship between borax concentration and RNA resolution, which can guide further experimental optimization.
