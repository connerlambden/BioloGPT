This notebook analyzes species composition data from green roofs, focusing on habitat connectivity and local environmental conditions.

```python
import pandas as pd

# Load species composition data
data = pd.read_csv('green_roof_data.csv')

# Analyze connectivity effects on species richness
connectivity_effects = data.groupby('connectivity').agg({'species_richness': 'mean'})
connectivity_effects
```

The analysis reveals how connectivity influences species richness on green roofs.

```python
# Visualize the results
import matplotlib.pyplot as plt

plt.bar(connectivity_effects.index, connectivity_effects['species_richness'])
plt.xlabel('Connectivity Level')
plt.ylabel('Average Species Richness')
plt.title('Impact of Connectivity on Species Richness')
plt.show()
```
