### Analyzing Microbial Communities in Symbiotic Plants
This notebook will explore the microbial community data associated with various symbiotic plants, focusing on OTU diversity.

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('symbiotic_plant_data.csv')

# Analyze OTU diversity
otu_summary = data.groupby('symbiotic_type').agg({'OTU_count': 'sum'})
print(otu_summary)
```
