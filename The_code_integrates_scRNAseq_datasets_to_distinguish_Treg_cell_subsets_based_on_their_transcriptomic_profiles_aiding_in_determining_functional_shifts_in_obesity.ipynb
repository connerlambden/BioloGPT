Analysis of single-cell RNA sequencing data to identify differential gene expression between tTreg and pTreg populations.

```python
import scanpy as sc
adata = sc.read_h5ad('VAT_Treg_cells.h5ad')
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.tl.pca(adata)
sc.pl.pca(adata, color='subset')
sc.tl.umap(adata)
sc.pl.umap(adata, color=['subset', 'Foxp3', 'IFN_gamma'])
```

This notebook provides a pipeline to visualize and interpret the shifts in Treg cell profiles with obesity.

```python
# Further comparative analysis
sc.tl.rank_genes_groups(adata, 'subset', method='wilcoxon')
sc.pl.rank_genes_groups(adata, n_genes=20, sharey=False)
```
