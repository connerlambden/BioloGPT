Loading the extracted dose-response data and plotting both iTR and MR against ivermectin intake.

```python
import plotly.graph_objects as go
import pandas as pd

data = pd.DataFrame([
    {'Treatment': 'Control', 'IVM_Intake': 0, 'iTR': 78.47, 'MR': 2.96},
    {'Treatment': 'T1', 'IVM_Intake': 0.019, 'iTR': 69.32, 'MR': 3.00},
    {'Treatment': 'T10', 'IVM_Intake': 0.196, 'iTR': 60.80, 'MR': 2.56},
    {'Treatment': 'T100', 'IVM_Intake': 0.409, 'iTR': 17.65, 'MR': 1.46},
    {'Treatment': 'T1000', 'IVM_Intake': 1.212, 'iTR': 8.44, 'MR': 0.63}
])

fig = go.Figure()
fig.add_trace(go.Scatter(x=data['IVM_Intake'], y=data['iTR'], mode='lines+markers', name='iTR (°C min-1)', line=dict(color='#6A0C76')))
fig.add_trace(go.Scatter(x=data['IVM_Intake'], y=data['MR'], mode='lines+markers', name='MR (W g-1)', line=dict(color='#FF5733')))

fig.update_layout(title='Ivermectin Dose-Response Effects on Dung Beetles', xaxis_title='IVM Intake (µg g-1)', yaxis_title='Measured Value', template='plotly_white')
fig.show()
```

This graph clearly illustrates the decline in both iTR and MR with increasing ivermectin concentrations.
