### Growth Performance Analysis of Methionine Sources in Poultry

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for growth performance
data = {
    'Diet': ['Basal', 'L-Met 0.3%', 'HMTBa 0.3%'],
    'SGR': [1.07, 1.70, 1.60],
    'FE': [0.71, 0.95, 0.89],
    'PER': [1.47, 1.97, 1.87]
}

growth_df = pd.DataFrame(data)

# Plotting the growth performance
plt.figure(figsize=(10, 6))
plt.bar(growth_df['Diet'], growth_df['SGR'], color='blue', alpha=0.6, label='Specific Growth Rate (SGR)')
plt.bar(growth_df['Diet'], growth_df['FE'], color='green', alpha=0.6, label='Feed Efficiency (FE)')
plt.bar(growth_df['Diet'], growth_df['PER'], color='orange', alpha=0.6, label='Protein Efficiency Ratio (PER)')
plt.title('Growth Performance of Poultry on Different Methionine Sources')
plt.xlabel('Diet')
plt.ylabel('Performance Metrics')
plt.legend()
plt.show()
```

### Discussion
This analysis provides a visual representation of the growth performance metrics for poultry fed different methionine sources.
