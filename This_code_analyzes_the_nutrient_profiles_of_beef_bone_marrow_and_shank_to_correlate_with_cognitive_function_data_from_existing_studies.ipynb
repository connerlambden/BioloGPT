### Nutrient Analysis of Beef Bone Marrow and Shank

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Nutrient data for beef bone marrow and shank
nutrient_data = {
    'Nutrient': ['Omega-3', 'Vitamin B12', 'Iron', 'Zinc'],
    'Bone Marrow': [5.0, 2.0, 1.5, 0.5],  # values in grams
    'Beef Shank': [2.0, 1.5, 3.0, 2.0]
}

# Create DataFrame
df = pd.DataFrame(nutrient_data)

# Plotting
plt.figure(figsize=(10, 6))
for nutrient in df['Nutrient']:
    plt.plot(df['Nutrient'], df[nutrient], marker='o', label=nutrient)
plt.title('Nutrient Comparison: Beef Bone Marrow vs. Beef Shank')
plt.xlabel('Nutrients')
plt.ylabel('Amount (g)')
plt.legend()
plt.grid()
plt.show()
```
