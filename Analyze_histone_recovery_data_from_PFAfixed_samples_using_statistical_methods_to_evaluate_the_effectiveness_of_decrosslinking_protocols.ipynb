This notebook will analyze histone recovery data, comparing samples treated with and without decrosslinking.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load histone recovery data
recovery_data = pd.read_csv('histone_recovery_data.csv')

# Analyze recovery rates
plt.figure(figsize=(10, 6))
plt.bar(recovery_data['Sample'], recovery_data['Recovery_Rate'])
plt.title('Histone Recovery Rates')
plt.xlabel('Sample Type')
plt.ylabel('Recovery Rate (%)')
plt.show()
```
