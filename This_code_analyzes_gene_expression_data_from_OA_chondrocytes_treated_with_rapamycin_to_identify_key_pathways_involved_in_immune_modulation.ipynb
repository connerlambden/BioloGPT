### Gene Expression Analysis of OA Chondrocytes Treated with Rapamycin

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('rapamycin_OA_chondrocytes.csv')
sns.boxplot(x='treatment', y='gene_expression', data=data)
plt.title('Gene Expression in OA Chondrocytes')
plt.show()
```

This analysis visualizes the impact of rapamycin on gene expression in OA chondrocytes, highlighting its potential effects on immune modulation.
