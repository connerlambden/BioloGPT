This notebook downloads longitudinal absenteeism data and applies statistical analyses to visualize trends and detect significant differences pre- and post-pandemic.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load dataset
absenteeism_data = pd.read_csv('absenteeism_data.csv')
# Plot trend
plt.figure(figsize=(10,6))
plt.plot(absenteeism_data['Year'], absenteeism_data['Sick_Leave_Days'], marker='o', color='#6A0C76')
plt.title('Absenteeism Trend (2017-2023)')
plt.xlabel('Year')
plt.ylabel('Sick Leave Days')
plt.grid(True)
plt.show()
```

The code generates a time-series plot using Matplotlib, which can be further expanded with regression analysis to validate the trend of sick leave days over the study duration.

```python
# Example regression analysis
import statsmodels.api as sm
X = absenteeism_data['Year']
X = sm.add_constant(X)
Y = absenteeism_data['Sick_Leave_Days']
model = sm.OLS(Y, X).fit()
print(model.summary())
```
