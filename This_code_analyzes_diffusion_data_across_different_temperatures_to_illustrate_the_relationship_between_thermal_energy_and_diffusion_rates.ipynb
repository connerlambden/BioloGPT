### Diffusion Rate Analysis
This notebook will analyze diffusion rates at varying temperatures using simulated data.

```python
import pandas as pd
import matplotlib.pyplot as plt

diffusion_data = pd.DataFrame({
    'Temperature (C)': [0, 10, 20, 30, 40],
    'Diffusion Rate (cm^2/s)': [0.1, 0.2, 0.5, 1.0, 1.5]
})

plt.plot(diffusion_data['Temperature (C)'], diffusion_data['Diffusion Rate (cm^2/s)'], marker='o')
plt.title('Diffusion Rate vs Temperature')
plt.xlabel('Temperature (C)')
plt.ylabel('Diffusion Rate (cm^2/s)')
plt.grid()
plt.show()
```
