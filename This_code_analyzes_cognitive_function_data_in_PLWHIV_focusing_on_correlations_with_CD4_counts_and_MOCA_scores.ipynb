### Data Analysis of Cognitive Function in PLWHIV
This notebook analyzes the relationship between CD4 counts and cognitive function as measured by MOCA scores.

```python
import pandas as pd

# Sample data
data = {
    'CD4_Count': [200, 300, 400, 500, 600],
    'MOCA_Score': [22, 24, 25, 26, 27]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate correlation
correlation = df['CD4_Count'].corr(df['MOCA_Score'])
correlation
```

### Correlation Analysis
The correlation coefficient between CD4 counts and MOCA scores indicates the strength of the relationship.

```python
# Display correlation result
print(f'Correlation between CD4 Count and MOCA Score: {correlation}')
```
