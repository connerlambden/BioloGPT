# Analyzing Atmospheric Gas Compositions Over Geological Time
This notebook will explore the changes in atmospheric gas compositions from early Earth to present.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load atmospheric data
# Example data structure: {'Time (billion years ago)': [4.5, 3.5, 2.5, 0], 'Oxygen (%)': [0, 0, 21, 21]}
data = {'Time (billion years ago)': [4.5, 3.5, 2.5, 0], 'Oxygen (%)': [0, 0, 21, 21]}
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(df['Time (billion years ago)'], df['Oxygen (%)'], marker='o')
plt.title('Change in Atmospheric Oxygen Levels Over Time')
plt.xlabel('Time (billion years ago)')
plt.ylabel('Oxygen (%)')
plt.grid()
plt.show()
```
