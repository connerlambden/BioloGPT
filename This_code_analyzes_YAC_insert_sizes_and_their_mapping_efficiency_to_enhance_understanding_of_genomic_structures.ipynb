### YAC Insert Size Analysis
This notebook analyzes the insert sizes of various YACs and their mapping efficiency to specific genomic regions.

```python
import pandas as pd

# Sample data for YAC insert sizes and mapping efficiency
data = {
    'YAC_clone': ['RS46', 'RS47', 'RS48'],
    'insert_size_kb': [70.4, 80.0, 90.5],
    'mapped_region': ['Xq27.3', 'Xq28.1', 'Xq29.2'],
    'mapping_efficiency': [95, 90, 85]
}
df = pd.DataFrame(data)

# Display the DataFrame
df
```

### Analysis of Mapping Efficiency
This section will analyze the mapping efficiency of YACs to understand their effectiveness in genomic studies.

```python
import matplotlib.pyplot as plt

# Plotting the mapping efficiency
plt.bar(df['YAC_clone'], df['mapping_efficiency'], color='blue')
plt.xlabel('YAC Clone')
plt.ylabel('Mapping Efficiency (%)')
plt.title('Mapping Efficiency of YAC Clones')
plt.show()
```
