## Genotype Frequency Analysis
This notebook analyzes the frequency changes of the HB and HS supergene alleles over multiple generations to understand their impact on population dynamics.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('genotype_frequencies.csv')
plt.plot(data['Generation'], data['HB_freq'], label='HB')
plt.plot(data['Generation'], data['HS_freq'], label='HS')
plt.xlabel('Generation')
plt.ylabel('Frequency')
plt.title('Supergene Allele Frequencies Over Generations')
plt.legend()
plt.show()
```
