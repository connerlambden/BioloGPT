Collect and process biofilm CFU log reduction data from the study for MRSA, Pseudomonas aeruginosa, and E. coli.

```python
import pandas as pd
import plotly.express as px

# Data from the study
 data = {'Bacterial Strain': ['MRSA', 'Pseudomonas aeruginosa', 'E. coli'],
         'aBL_LR': [-1.76, -3.38, -1.93],
         'aBL_TC_LR': [-2.72, -5.39, -3.82]}
df = pd.DataFrame(data)

# Create a grouped bar chart
fig = px.bar(df, x='Bacterial Strain', y=['aBL_LR','aBL_TC_LR'],
             title='Log Reduction Comparison: aBL vs aBL + Tetracycline',
             labels={'value':'Log Reduction','variable':'Treatment'},
             barmode='group',
             color_discrete_sequence=['#6A0C76','#FFA500'])
fig.show()
```
