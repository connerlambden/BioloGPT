This notebook will analyze skin microbiota sequencing data to identify microbial species that correlate with healthy and diseased skin states.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load skin microbiota dataset
data = pd.read_csv('skin_microbiota_data.csv')

# Analyze microbial diversity
sns.boxplot(x='health_status', y='diversity_index', data=data)
plt.title('Microbial Diversity by Health Status')
plt.show()
```

The analysis will visualize differences in microbial diversity between healthy and diseased skin.

```python
# Identify key species
key_species = data.groupby('health_status')['species'].value_counts().unstack().fillna(0)
key_species.plot(kind='bar', stacked=True)
plt.title('Key Microbial Species by Health Status')
plt.show()
```

This plot will help identify which microbial species are predominant in healthy versus diseased skin.
