### Impedance Data Analysis
This notebook analyzes impedance measurements to understand monocyte adhesion effects on macrophage differentiation.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load impedance data
impedance_data = pd.read_csv('impedance_data.csv')

# Analyze correlation between adhesion and differentiation
correlation = impedance_data.corr()
plt.figure(figsize=(10, 6))
plt.title('Correlation between Impedance and Macrophage Differentiation')
plt.imshow(correlation, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.show()
```
