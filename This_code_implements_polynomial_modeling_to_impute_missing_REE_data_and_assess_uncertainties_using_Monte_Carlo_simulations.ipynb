### Polynomial Modeling for REE Data Imputation
This notebook demonstrates how to use polynomial modeling to impute missing REE data and quantify uncertainties.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Load REE dataset
ree_data = pd.read_csv('ree_data.csv')

# Polynomial feature transformation
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(ree_data[['REE1', 'REE2', 'REE3']])

# Fit model
model = LinearRegression()
model.fit(X_poly, ree_data['Anomaly'])

# Predict missing values
predictions = model.predict(X_poly)
```
