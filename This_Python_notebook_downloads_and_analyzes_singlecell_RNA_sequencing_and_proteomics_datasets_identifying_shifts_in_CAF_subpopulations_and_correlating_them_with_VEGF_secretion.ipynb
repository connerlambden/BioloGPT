Import single-cell RNA-seq datasets of bCAFs to identify phenotype shifts with MCL-1 inhibition.

```python
import scanpy as sc
adata = sc.read('data/bCAFs_scRNAseq.h5ad')
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.tl.pca(adata)
sc.pl.pca(adata, color=['MCL1', 'VEGFA'])
```

Analyze proteomics data linking MCL-1 inhibition to mitochondrial fission markers and VEGF levels.

```python
import pandas as pd
proteomics = pd.read_csv('data/proteomics.csv')
print(proteomics[['MCL1', 'DRP1', 'VEGFA']].describe())
```
