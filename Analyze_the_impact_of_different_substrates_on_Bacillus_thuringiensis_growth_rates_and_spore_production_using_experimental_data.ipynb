This notebook analyzes the effects of various substrates on Bacillus thuringiensis growth.

```python
import pandas as pd
# Load experimental data
experimental_data = pd.read_csv('substrate_experiment_data.csv')
```

The data includes growth rates and spore counts for different substrates.

```python
# Analyze growth rates
growth_analysis = experimental_data.groupby('substrate').mean()
```

This analysis will help identify the most effective substrates for Bacillus thuringiensis cultivation.

```python
growth_analysis.plot(kind='bar', title='Average Growth Rates by Substrate')
```
