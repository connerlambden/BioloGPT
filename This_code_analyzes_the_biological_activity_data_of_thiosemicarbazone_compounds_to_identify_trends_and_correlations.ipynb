### Analysis of Thiosemicarbazone Compounds
This notebook will analyze the biological activity data of thiosemicarbazone compounds, focusing on their antibacterial, antioxidant, and anticancer properties.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for thiosemicarbazone compounds
activity_data = {
    'Compound': ['L1', 'L', 'L4'],
    'MIC (mg/L)': [10, None, None],
    'IC50 (ABTS mg/mL)': [None, 0.010, None],
    'IC50 (DPPH mg/mL)': [None, 0.403, None],
    'IC50 (Cytotoxicity µM)': [None, None, 193.44]
}

# Create DataFrame
df = pd.DataFrame(activity_data)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['Compound'], df['MIC (mg/L)'], label='MIC (mg/L)', color='blue')
plt.bar(df['Compound'], df['IC50 (ABTS mg/mL)'], label='IC50 (ABTS mg/mL)', color='orange')
plt.bar(df['Compound'], df['IC50 (DPPH mg/mL)'], label='IC50 (DPPH mg/mL)', color='green')
plt.bar(df['Compound'], df['IC50 (Cytotoxicity µM)'], label='IC50 (Cytotoxicity µM)', color='red')
plt.title('Biological Activity of Thiosemicarbazone Compounds')
plt.xlabel('Compounds')
plt.ylabel('Activity')
plt.legend()
plt.show()
```
