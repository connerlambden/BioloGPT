Analisi dei dati di espressione genica in relazione alle modifiche degli istoni.

```python
import pandas as pd
# Carica i dati di espressione genica e modifiche degli istoni
gene_expression_data = pd.read_csv('gene_expression.csv')
histone_modifications_data = pd.read_csv('histone_modifications.csv')
```

Unisci i dataset e analizza le correlazioni.

```python
merged_data = pd.merge(gene_expression_data, histone_modifications_data, on='gene_id')
correlation = merged_data.corr()
```

Visualizza i risultati.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(correlation, annot=True)
plt.title('Correlazione tra espressione genica e modifiche degli istoni')
plt.show()
```
