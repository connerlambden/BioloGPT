جمع البيانات الزراعية من حماه وتحليلها لفهم تأثير التغير المناخي.

```python
import pandas as pd

# تحميل البيانات الزراعية
data = pd.read_csv('hama_agriculture_data.csv')

# تحليل البيانات
summary = data.describe()
```
