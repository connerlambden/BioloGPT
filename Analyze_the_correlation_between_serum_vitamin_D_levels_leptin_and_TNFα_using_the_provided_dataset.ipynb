### Correlation Analysis of Vitamin D, Leptin, and TNF-α
This notebook analyzes the relationships between serum vitamin D levels, leptin, and TNF-α in overweight postmenopausal women with hypertension.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data as provided
data = {
    'Group': ['NW-NT']*116 + ['NW-HT']*115 + ['OW-HT']*115,
    'Age': [55.32]*116 + [55.33]*115 + [55.35]*115,
    'BMI': [23.46]*116 + [23.42]*115 + [27.49]*115,
    'Leptin': [5.99]*116 + [10.23]*115 + [13.40]*115,
    '25(OH)D': [32.04]*116 + [32.05]*115 + [30.21]*115,
    'IL-6': [7.68]*116 + [9.26]*115 + [13.69]*115,
    'TNF-α': [6.01]*116 + [9.23]*115 + [12.61]*115
}

df = pd.DataFrame(data)

# Calculate correlations
correlations = df[['BMI', 'Leptin', '25(OH)D', 'IL-6', 'TNF-α']].corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlations, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

### Interpretation
The correlation matrix reveals significant negative correlations between vitamin D levels and both leptin and TNF-α, as well as positive correlations between BMI and these inflammatory markers.
