### Gene Expression Analysis for Metabolic Syndrome and COVID-19 Severity

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_data.csv')
sns.boxplot(x='Condition', y='Gene_Expression', data=data)
plt.title('Gene Expression in Metabolic Syndrome vs Control')
plt.show()
```
