Load the zebrafish single-cell RNA-seq dataset and filter for cardiomyocyte clusters to analyze the expression profiles of ttn.2 positive and negative cells.

```python
import scanpy as sc
adata = sc.read_h5ad('zebrafish_heart_cells.h5ad')
# Filter for cardiomyocytes
cm = adata[adata.obs['cell_type'] == 'cardiomyocyte']
# Create a mask for ttn.2 positive cells
cm.obs['ttn2_pos'] = cm[:, 'ttn.2'].X.flatten() > 0
sc.pl.umap(cm, color='ttn2_pos', title='ttn.2 Positive vs Negative Cardiomyocytes')
```

The code visualizes the UMAP distribution of ttn.2 positive and negative cardiomyocytes.
