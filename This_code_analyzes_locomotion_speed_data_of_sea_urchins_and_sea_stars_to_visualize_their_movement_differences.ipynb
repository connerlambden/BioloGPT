### Locomotion Speed Analysis of Sea Urchins and Sea Stars

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for sea urchins and sea stars
data = {'Species': ['Sea Urchin', 'Sea Star'], 'Speed (cm/min)': [50, 12]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Species'], df['Speed (cm/min)'], color=['blue', 'orange'])
plt.title('Locomotion Speed of Sea Urchins vs Sea Stars')
plt.ylabel('Speed (cm/min)')
plt.show()
```

This analysis provides a visual representation of the speed differences between sea urchins and sea stars.
