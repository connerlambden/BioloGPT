### Step-by-Step Guide to Generate a Punnett Square
1. Define the alleles of both parents.
2. Create a grid to represent the combinations of these alleles.
3. Fill in the grid with the resulting genotypes.

```python
def punnett_square(parent1, parent2):
    square = [[f'{a1}{a2}' for a2 in parent2] for a1 in parent1]
    return square

# Example usage
parent1 = ['A', 'a']
parent2 = ['A', 'a']
print(punnett_square(parent1, parent2))
```
