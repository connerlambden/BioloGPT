Análisis de datos sobre la gestión de residuos químicos en laboratorios.

```python
import pandas as pd
# Cargar datos de gestión de residuos
data = pd.read_csv('residuos_quimicos.csv')
# Análisis de residuos generados por tipo de laboratorio
residuos_por_tipo = data.groupby('tipo_laboratorio').sum()
```

Este análisis permite identificar patrones en la generación de residuos y optimizar su gestión.

```python
residuos_por_tipo.to_csv('residuos_analizados.csv')
```
