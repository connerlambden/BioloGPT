### Step 1: Data Preparation
Load the relevant datasets from the studies on blood microbiome and transcriptomics.

```python
import pandas as pd

# Load datasets
dataset_microbiome = pd.read_csv('microbiome_data.csv')
dataset_transcriptomics = pd.read_csv('transcriptomics_data.csv')
```

### Step 2: Data Integration
Merge the microbiome and transcriptomic datasets based on patient identifiers.

```python
# Merge datasets
merged_data = pd.merge(dataset_microbiome, dataset_transcriptomics, on='patient_id')
```

### Step 3: Feature Selection
Identify significant features that differentiate TB patients from healthy controls.

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Select top features
X = merged_data.drop('diagnosis', axis=1)
y = merged_data['diagnosis']
selector = SelectKBest(score_func=f_classif, k=10)
X_new = selector.fit_transform(X, y)
```
