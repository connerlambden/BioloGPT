This notebook retrieves species count data and uses Plotly for bar chart visualization to compare mosquito distribution across zones.

```python
import plotly.graph_objs as go

# Data
n_gambiae = [1042, 898, 396]
n_funestus = [480, 86, 109]
n_coustani = [216, 34, 13]
n_pharoensis = [3, 0, 2]
n_maculipalpis = [53, 0, 10]
n_leesoni = [0, 0, 0]
zones = ['Urban', 'Peri-urban', 'Rural']

# Bar chart traces
trace1 = go.Bar(x=zones, y=n_gambiae, name='An. gambiae s.l.')
trace2 = go.Bar(x=zones, y=n_funestus, name='An. funestus s.l.')
trace3 = go.Bar(x=zones, y=n_coustani, name='An. coustani')
trace4 = go.Bar(x=zones, y=n_pharoensis, name='An. pharoensis')
trace5 = go.Bar(x=zones, y=n_maculipalpis, name='An. maculipalpis')
trace6 = go.Bar(x=zones, y=n_leesoni, name='An. leesoni')

data = [trace1, trace2, trace3, trace4, trace5, trace6]
layout = go.Layout(title='Mosquito Species Distribution in Kisumu Zones', barmode='group')

fig = go.Figure(data=data, layout=layout)
fig.show()
```

The above code generates a grouped bar chart that visually compares the species counts across the three zones.
