This notebook downloads the enzyme activity dataset and generates bar charts to visualize both enzyme activities and kinetic parameters.

```python
import pandas as pd
import plotly.express as px

data = {
    'Parameter': ['Cellulase Activity (U/mL)', 'Exo-β-1,4-glucanase Activity (U/mL)', 'β-1,4-glucanase Activity (U/mL)', 'K m (mM)', 'V max (U/mg)'],
    'Value': [0.079, 0.007, 0.005, 3.13, 0.16]
}
df = pd.DataFrame(data)

fig = px.bar(df, x='Parameter', y='Value', title='Enzyme Activity and Kinetic Parameters')
fig.show()

print(df.describe())
```
