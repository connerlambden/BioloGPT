### Analyzing Student Performance Data
This notebook will analyze the exam performance data of students who utilized online databases.

```python
import pandas as pd

# Sample data
exam_data = {
    'Year': [2012, 2013, 2014, 2015],
    'Exam_1': [88.7, 89.1, 95.3, 104.9],
    'Exam_2': [98.3, 94.0, 105.7, 83.6]
}

df = pd.DataFrame(exam_data)

# Calculate mean scores
mean_scores = df[['Exam_1', 'Exam_2']].mean()
mean_scores
```
