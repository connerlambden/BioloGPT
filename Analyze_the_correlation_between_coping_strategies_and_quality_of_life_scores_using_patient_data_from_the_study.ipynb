This notebook analyzes the relationship between coping strategies and quality of life scores among breast cancer patients in Iraq.

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('breast_cancer_qol_data.csv')

# Analyze the correlation between coping strategies and QOL scores
correlation = data[['passive_coping', 'active_coping', 'global_qol']].corr()
```

The correlation matrix provides insights into how different coping strategies relate to overall quality of life.

```python
print(correlation)
```
