### Data Analysis of Dorsal Scute Microchemistry
This notebook will analyze the microchemical data of dorsal scutes to identify habitat transitions in Atlantic Sturgeon.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load microchemistry data
data = pd.read_csv('dorsal_scute_microchemistry.csv')

# Analyze the relationship between Sr:Ca ratios and habitat types
plt.figure(figsize=(10, 6))
plt.scatter(data['Sr_Ca'], data['Habitat_Type'])
plt.title('Sr:Ca Ratios vs Habitat Type')
plt.xlabel('Sr:Ca Ratio')
plt.ylabel('Habitat Type')
plt.show()
```
