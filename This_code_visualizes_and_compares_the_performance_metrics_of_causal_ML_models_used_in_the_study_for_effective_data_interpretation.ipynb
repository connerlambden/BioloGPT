Below, we load the extracted performance metrics of the causal ML models from the study data to generate a table and bar chart using Plotly.

```python
import plotly.graph_objects as go
import pandas as pd

# Create a DataFrame with extracted model performance
data = pd.DataFrame({
    'Model': ['Causal KNN', 'Causal Random Forest'],
    'Accuracy': [94.52, 90.3],
    'Precision': [90.25, 92.2],
    'Recall': [92.57, 94.3],
    'F1 Score': [88.2, 89.2]
})

# Create and display a table summarizing the performance metrics
fig_table = go.Figure(data=[go.Table(
    header=dict(
        values=[list(data.columns)],
        fill_color='#6A0C76',
        font=dict(color='white', size=12),
        align='center'
    ),
    cells=dict(
        values=[data[k] for k in data.columns],
        fill_color='lavender',
        font=dict(color='black', size=11),
        align='center'
    )
)])
fig_table.show()
```

The table above details the accuracy, precision, recall, and F1 Score for each of the evaluated causal models.
