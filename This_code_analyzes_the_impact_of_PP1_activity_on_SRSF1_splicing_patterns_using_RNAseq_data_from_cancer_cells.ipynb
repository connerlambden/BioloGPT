This notebook will analyze RNA-seq data to assess the splicing patterns of SRSF1 in response to PP1 modulation.

```python
import pandas as pd
import numpy as np

# Load RNA-seq data
rna_seq_data = pd.read_csv('path_to_rna_seq_data.csv')

# Analyze splicing events
splicing_analysis = rna_seq_data.groupby('gene').agg({'exon_inclusion': 'mean'})

# Output results
splicing_analysis.to_csv('splicing_analysis_results.csv')
```

The results will provide insights into how PP1 activity influences SRSF1 splicing in cancer cells.

```python
# Visualize results
import matplotlib.pyplot as plt
plt.bar(splicing_analysis.index, splicing_analysis['exon_inclusion'])
plt.title('SRSF1 Exon Inclusion Rates')
plt.xlabel('Gene')
plt.ylabel('Exon Inclusion Rate')
plt.show()
```
