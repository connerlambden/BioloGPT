### Data Analysis of Atherosclerotic Disease Risk Factors
This analysis focuses on identifying key risk factors associated with severe atherosclerotic disease in patients.

```python
import pandas as pd

# Load patient data
patient_data = pd.read_csv('patient_data.csv')

# Analyze risk factors
risk_factors = patient_data[['age', 'diabetes_duration', 'hypertension', 'cholesterol']].describe()
```

### Summary of Risk Factors
The analysis provides insights into the average age, diabetes duration, and prevalence of hypertension and high cholesterol among patients.

```python
risk_factors.to_csv('risk_factors_summary.csv')
```
