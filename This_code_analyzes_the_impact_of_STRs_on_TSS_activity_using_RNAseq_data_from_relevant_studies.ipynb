### Data Preparation
Load RNA-seq data and STR information from relevant studies.

```python
import pandas as pd

# Load datasets
df_str = pd.read_csv('path_to_str_data.csv')
df_rna_seq = pd.read_csv('path_to_rna_seq_data.csv')
```

### Data Analysis
Merge datasets and analyze the correlation between STRs and TSS activity.

```python
merged_df = pd.merge(df_str, df_rna_seq, on='gene_id')
correlation = merged_df.corr()['str_activity']['tss_activity']
print(f'Correlation between STRs and TSS activity: {correlation}')
```

### Visualization
Visualize the relationship between STRs and TSS activity.

```python
import matplotlib.pyplot as plt

plt.scatter(merged_df['str_activity'], merged_df['tss_activity'])
plt.title('STR Activity vs TSS Activity')
plt.xlabel('STR Activity')
plt.ylabel('TSS Activity')
plt.show()
```
