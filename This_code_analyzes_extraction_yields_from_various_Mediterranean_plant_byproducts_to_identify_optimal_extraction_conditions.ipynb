# Analysis of Extraction Yields from Mediterranean Plant By-Products

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Plant': ['Pistacia lentiscus', 'Grape Pomace', 'Olive Leaves', 'Tomato Peels'],
        'Yield (mg/g)': [25.4, 30.2, 22.5, 18.0]}
df = pd.DataFrame(data)

plt.bar(df['Plant'], df['Yield (mg/g)'])
plt.title('Extraction Yields of Anthocyanins and Flavonols')
plt.xlabel('Plant')
plt.ylabel('Yield (mg/g)')
plt.show()
```

This analysis visualizes the extraction yields of anthocyanins and flavonols from various Mediterranean plant by-products.
