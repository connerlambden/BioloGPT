This notebook analyzes solubility data for various compounds in water, focusing on the effects of polarity and temperature.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load solubility data
solubility_data = pd.read_csv('solubility_data.csv')

# Analyze trends
plt.figure(figsize=(10, 6))
plt.scatter(solubility_data['Polarity'], solubility_data['Solubility'])
plt.title('Solubility vs Polarity')
plt.xlabel('Polarity')
plt.ylabel('Solubility')
plt.show()
```

This analysis provides insights into how polarity affects solubility in water.
