This notebook simulates the diffusion of MinD on membranes with varying curvature, providing insights into how curvature affects diffusion rates.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
curvature = np.linspace(0, 1, 100)  # Curvature values
D_base = 0.1  # Base diffusion coefficient
D_curvature_effect = D_base * (1 - curvature)  # Effect of curvature on diffusion

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(curvature, D_curvature_effect, label='Diffusion Coefficient', color='blue')
plt.title('Effect of Membrane Curvature on MinD Diffusion')
plt.xlabel('Curvature')
plt.ylabel('Diffusion Coefficient (µm²/s)')
plt.legend()
plt.grid()
plt.show()
```

This simulation illustrates how increasing curvature reduces the diffusion coefficient of MinD, highlighting the quantitative relationship between curvature and diffusion.
