This notebook will analyze the distribution of Okazaki fragments and their impact on nucleosome positioning in rad27 mutants.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load sequencing data
okazaki_data = pd.read_csv('okazaki_fragments.csv')

# Analyze distribution
plt.hist(okazaki_data['fragment_length'], bins=50)
plt.title('Distribution of Okazaki Fragment Lengths')
plt.xlabel('Fragment Length (bp)')
plt.ylabel('Frequency')
plt.show()
```
