Load and preprocess the dataset containing metabolic and resistance data for all 66 bacterial strains.

```python
import pandas as pd
import plotly.express as px

# Download dataset from provided Zenodo link
url = 'https://zenodo.org/records/10954779'
data = pd.read_csv(url, sep='\t')

# Preview data
print(data.head())
```

Generate visualizations such as bar plots and heatmaps to explore the relationship between metabolic activity and antagonistic behavior across organs.

```python
fig = px.bar(data, x='Organ', y='Metabolic_Activity', color='Organ', title='Metabolic Activity by Organ')
fig.show()
```

Perform clustering analysis on antibiotic resistance profiles to validate organ-specific grouping.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a heatmap for the resistance data
pivot_data = data.pivot('Strain_ID', 'Antibiotic', 'Resistance_Score')
sns.heatmap(pivot_data, cmap='viridis')
plt.title('Antibiotic Resistance Heatmap')
plt.show()
```
