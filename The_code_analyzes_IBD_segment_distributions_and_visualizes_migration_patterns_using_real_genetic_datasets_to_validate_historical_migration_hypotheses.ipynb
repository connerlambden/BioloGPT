This notebook downloads the real genetic datasets from the Danish and UK biobanks, processes IBD segment data using FastSMC outputs, and visualizes regional genetic connectivity.

```python
import pandas as pd
import plotly.express as px

# Assume df is a DataFrame loaded from the real biobank IBD dataset
# df should contain columns 'region', 'IBD_length', 'IBD_count'

df = pd.read_csv('real_ibd_data.csv')
fig = px.scatter(df, x='IBD_length', y='IBD_count', color='region', title='Regional IBD Sharing Patterns')
fig.show()
```

The above code provides a visualization of the IBD sharing patterns across different regions, helping to further validate the migration signals inferred in the study.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
ax = sns.boxplot(data=df, x='region', y='IBD_length')
ax.set_title('Distribution of IBD Segment Lengths Across Regions')
plt.show()
```
