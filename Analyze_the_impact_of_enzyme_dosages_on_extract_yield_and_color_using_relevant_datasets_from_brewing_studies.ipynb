This notebook will analyze the effects of enzyme dosages on extract yield and color in lentil-based wort using empirical data.

```python
import pandas as pd

# Load relevant datasets
dataset = pd.read_csv('path_to_lentil_brewing_data.csv')

# Analyze the impact of enzyme dosages on extract yield and color
results = dataset.groupby('enzyme_dosage').agg({'extract_yield': 'mean', 'color': 'mean'})
results.plot(kind='bar')
```

The analysis will provide insights into how varying enzyme dosages affect the quality parameters of lentil-based wort.
