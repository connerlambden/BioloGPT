### Data Analysis of MSSA Bacteremia and Endocarditis Incidence

```python
import pandas as pd

# Load patient data
patient_data = pd.read_csv('mssa_bacteremia_data.csv')

# Analyze endocarditis incidence
endocarditis_incidence = patient_data['endocarditis'].mean()
print(f'Endocarditis incidence: {endocarditis_incidence * 100:.2f}%')
```
