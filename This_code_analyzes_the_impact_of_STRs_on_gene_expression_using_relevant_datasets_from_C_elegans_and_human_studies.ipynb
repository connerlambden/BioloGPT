### Data Preparation
Load the necessary libraries and datasets for analysis.

```python
import pandas as pd
import numpy as np

# Load datasets
str_data = pd.read_csv('path_to_str_data.csv')
gene_expression_data = pd.read_csv('path_to_gene_expression_data.csv')
```

### Data Analysis
Perform statistical analysis to identify significant associations between STRs and gene expression.

```python
from scipy import stats

# Example analysis: Correlation between STR length and gene expression
correlation_results = stats.pearsonr(str_data['STR_length'], gene_expression_data['expression_level'])
print('Correlation coefficient:', correlation_results[0])
print('P-value:', correlation_results[1])
```

### Results Visualization
Visualize the results using appropriate plots.

```python
import matplotlib.pyplot as plt

plt.scatter(str_data['STR_length'], gene_expression_data['expression_level'])
plt.title('STR Length vs Gene Expression')
plt.xlabel('STR Length')
plt.ylabel('Gene Expression Level')
plt.show()
```
