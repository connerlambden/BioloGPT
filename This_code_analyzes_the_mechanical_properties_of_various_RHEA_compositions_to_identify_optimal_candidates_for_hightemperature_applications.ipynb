This notebook will analyze the mechanical properties of different RHEA compositions, focusing on yield strength and ductility.

```python
import pandas as pd

# Sample data for RHEA compositions and their properties
data = {
    'Composition': ['Mo72.3W12.8Ta10.0Ti2.5Zr2.5', 'MoNbTaW'],
    'Room_Temperature_Yield_Strength': [820, 1010],
    'High_Temperature_Yield_Strength': [446, 460]
}
df = pd.DataFrame(data)

# Analyzing the data
print(df.describe() )
```

The analysis provides insights into the mechanical properties of the selected RHEA compositions.

```python
# Further analysis can be conducted to visualize the data
import matplotlib.pyplot as plt

plt.bar(df['Composition'], df['Room_Temperature_Yield_Strength'], color='blue', label='Room Temperature Yield Strength')
plt.bar(df['Composition'], df['High_Temperature_Yield_Strength'], color='red', label='High Temperature Yield Strength', alpha=0.5)
plt.xlabel('Composition')
plt.ylabel('Yield Strength (MPa)')
plt.title('Mechanical Properties of RHEAs')
plt.legend()
plt.show()
```
