### Data Analysis of Neuronal Activity in Response to Sequential Visual Stimuli

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for PPC
data = {'Repeat Frequency': ['10 Hz', '24 Hz', '40 Hz'],
        'Mean PPC BW': [0.0879, 0.0251, 0.0012],
        'Mean PPC NW': [0.0870, 0.0138, 0.0055]}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Repeat Frequency'], df['Mean PPC BW'], label='Mean PPC BW', alpha=0.6)
plt.bar(df['Repeat Frequency'], df['Mean PPC NW'], label='Mean PPC NW', alpha=0.6)
plt.title('Pairwise Phase Consistency (PPC) Across Different Repeat Frequencies')
plt.xlabel('Repeat Frequency')
plt.ylabel('Mean PPC')
plt.legend()
plt.show()
```
