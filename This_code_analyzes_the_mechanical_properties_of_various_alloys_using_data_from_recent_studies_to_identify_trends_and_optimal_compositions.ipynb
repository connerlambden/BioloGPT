### Analysis of Alloy Properties
This notebook will analyze the mechanical properties of different alloys based on recent research data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for alloys
alloy_data = {
    'Alloy': ['Steel', 'Aluminum', 'Titanium', 'Magnesium-Titanium'],
    'Tensile Strength (MPa)': [400, 300, 900, 350],
    'Corrosion Resistance': [3, 5, 4, 2]  # Scale of 1-5
}

# Create DataFrame
alloy_df = pd.DataFrame(alloy_data)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.bar(alloy_df['Alloy'], alloy_df['Tensile Strength (MPa)'], color='blue', alpha=0.7)
plt.title('Tensile Strength of Various Alloys')
plt.xlabel('Alloy Type')
plt.ylabel('Tensile Strength (MPa)')
plt.show()
```

### Discussion
This analysis provides insights into the tensile strength and corrosion resistance of various alloys, highlighting their suitability for different applications.
