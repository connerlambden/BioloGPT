This notebook will calculate the melting temperature (Tm) for a set of primers using the NEB Tm Calculator API.

```python
import requests

# Define the primer sequences
forward_primer = "GAA CCT"
reverse_primer = "CAT CAC AAC CTG GAG CAT TG"

# Define the API endpoint
url = 'https://tmcalculator.neb.com/'

# Function to calculate Tm
def calculate_tm(primer):
    response = requests.post(url, data={'sequence': primer, 'salt_concentration': 50})
    return response.json()['Tm']

# Calculate Tm for both primers
forward_tm = calculate_tm(forward_primer)
reverse_tm = calculate_tm(reverse_primer)

forward_tm, reverse_tm
```

The Tm values for the forward and reverse primers have been calculated. Ensure to handle API responses and errors appropriately.

```python
# Display the results
print(f"Forward Primer Tm: {forward_tm}°C")
print(f"Reverse Primer Tm: {reverse_tm}°C")
```
