### Microbiome Analysis of Thoosa mismalolli
This notebook analyzes the operational taxonomic units (OTUs) associated with different life stages of the coral boring sponge Thoosa mismalolli.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for OTUs
data = {
    'Sample Type': ['Adult sponges', 'Brooding larvae', 'Early free-swimming larvae (inshore)', 'Free-swimming pelagic larvae (offshore to 1 km)', 'Free-swimming pelagic larvae (offshore to 3 km)', 'Seawater (inshore)', 'Seawater (offshore)'],
    'OTUs': [222, 178, 196, 403, 64, 647, 276]
}

df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10,6))
plt.bar(df['Sample Type'], df['OTUs'], color='skyblue')
plt.title('OTUs in Different Life Stages of Thoosa mismalolli')
plt.xlabel('Sample Type')
plt.ylabel('Number of OTUs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Discussion
This analysis provides insights into the diversity of the microbiome associated with Thoosa mismalolli, highlighting the importance of understanding these dynamics for marine ecology.
