### Oxygen Consumption Analysis
This notebook analyzes oxygen consumption data from adherent cells to identify metabolic shifts.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('oxygen_consumption_data.csv')
plt.plot(data['Time'], data['Oxygen_Consumption'])
plt.title('Oxygen Consumption Over Time')
plt.xlabel('Time (minutes)')
plt.ylabel('Oxygen Consumption (nmol/s)')
plt.show()
```
