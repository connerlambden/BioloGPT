Šis kods analizē enzīmu aktivitātes datus, izmantojot statistikas metodes.

```python
import pandas as pd

data = pd.read_csv('enzimu_dati.csv')
# Analizēt datus un noteikt aktivitāti
activity = data['activity'].mean()
```

Analīze parāda vidējo enzīmu aktivitāti, kas var palīdzēt izprast to efektivitāti.

```python
print('Vidējā aktivitāte:', activity)
```
