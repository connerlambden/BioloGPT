This notebook will analyze longitudinal data on AMR in NICUs, focusing on the relationship between antibiotic use and resistance patterns.

```python
import pandas as pd

# Load AMR data
data = pd.read_csv('amr_data.csv')

# Analyze trends over time
trends = data.groupby(['year', 'antibiotic']).agg({'resistance_rate': 'mean'}).reset_index()
```

The analysis will provide insights into how antibiotic use correlates with AMR trends over time.

```python
import matplotlib.pyplot as plt

# Plotting the trends
plt.figure(figsize=(10, 6))
for antibiotic in trends['antibiotic'].unique():
    subset = trends[trends['antibiotic'] == antibiotic]
    plt.plot(subset['year'], subset['resistance_rate'], label=antibiotic)
plt.title('AMR Trends Over Time by Antibiotic')
plt.xlabel('Year')
plt.ylabel('Resistance Rate')
plt.legend()
plt.show()
```
