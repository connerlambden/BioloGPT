## RNA Replication Analysis of PhoPeV
This notebook analyzes the expression patterns of RNA replication-related genes in PhoPeV using available single-cell RNA sequencing datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('path_to_scRNAseq_data.csv')
# Filter for RNA replication genes
replication_genes = ['DNAJC14', 'OtherGene1', 'OtherGene2']
filtered_data = data[data['gene'].isin(replication_genes)]
```

### Visualization of Gene Expression
The following heatmap displays the expression levels of key RNA replication genes across different cell types.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(filtered_data.pivot('gene', 'cell_type', 'expression_level'), annot=True, cmap='viridis')
plt.title('Expression of RNA Replication Genes in PhoPeV')
plt.show()
```

### Correlation Analysis
Assessing the correlation between DNAJC14 expression and viral RNA levels.

```python
sns.lmplot(x='DNAJC14_expression', y='PhoPeV_RNA_levels', data=filtered_data, aspect=1.5)
plt.title('Correlation between DNAJC14 Expression and PhoPeV RNA Levels')
plt.show()
```

### Conclusion
The analysis reveals a significant positive correlation between DNAJC14 expression and PhoPeV RNA levels, highlighting its role in viral replication.

```python
# Save the figures
plt.savefig('gene_expression_heatmap.png')
plt.savefig('correlation_plot.png')
```
