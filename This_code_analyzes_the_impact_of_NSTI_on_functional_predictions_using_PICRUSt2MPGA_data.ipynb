### Data Preparation
Load the relevant datasets for analysis.

```python
import pandas as pd

# Load the dataset
nsti_data = pd.read_csv('nsti_data.csv')

# Display the first few rows
nsti_data.head()
```

### Analysis of NSTI Impact
Calculate the correlation between NSTI values and functional predictions.

```python
# Calculate Spearman correlation
correlation = nsti_data['NSTI'].corr(nsti_data['Functional_Prediction'], method='spearman')
print(f'Spearman correlation: {correlation}')
```

### Visualization
Visualize the relationship between NSTI and functional predictions.

```python
import matplotlib.pyplot as plt

plt.scatter(nsti_data['NSTI'], nsti_data['Functional_Prediction'])
plt.title('NSTI vs Functional Prediction')
plt.xlabel('NSTI')
plt.ylabel('Functional Prediction')
plt.show()
```
