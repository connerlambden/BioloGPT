This notebook will explore gene expression profiles during T cell and B cell development.

```python
import pandas as pd
# Load relevant datasets
b_cell_data = pd.read_csv('b_cell_development.csv')
t_cell_data = pd.read_csv('t_cell_development.csv')
```

The following analysis will compare gene expression levels between T cells and B cells.

```python
# Example analysis code
b_cell_avg = b_cell_data.mean()
t_cell_avg = t_cell_data.mean()
comparison = pd.DataFrame({'B Cell': b_cell_avg, 'T Cell': t_cell_avg})
comparison.plot(kind='bar')
```
