### Data Analysis of SBT-AgNPs Degradation Efficiency

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Dye': ['Methylene Blue', 'Malachite Green'], 'Degradation (%)': [58.3, 60]}
df = pd.DataFrame(data)

plt.bar(df['Dye'], df['Degradation (%)'], color=['blue', 'green'])
plt.title('Degradation Efficiency of SBT-AgNPs')
plt.xlabel('Dye Type')
plt.ylabel('Degradation (%)')
plt.show()
```

This analysis visualizes the degradation efficiency of SBT-AgNPs on methylene blue and malachite green dyes.
