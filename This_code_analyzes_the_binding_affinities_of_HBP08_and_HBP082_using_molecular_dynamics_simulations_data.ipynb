### Binding Affinity Analysis of HBP08 and HBP08-2
This analysis will utilize the Kd and IC50 values obtained from experimental data to visualize the binding affinities of the peptides.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
peptides = ['HBP08', 'HBP08-2']
kd_values = [0.8, 0.0281]  # in µM
ic50_values = [50, 3.31]  # in µM

# Plotting
x = np.arange(len(peptides))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, kd_values, width, label='Kd (µM)')
rects2 = ax.bar(x + width/2, ic50_values, width, label='IC50 (µM)')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Concentration (µM)')
ax.set_title('Binding Affinity of HBP08 and HBP08-2')
ax.set_xticks(x)
ax.set_xticklabels(peptides)
ax.legend()

plt.show()
```
