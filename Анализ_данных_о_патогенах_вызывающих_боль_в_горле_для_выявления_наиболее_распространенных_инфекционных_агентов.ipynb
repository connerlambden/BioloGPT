Анализ данных о вирусах и бактериях, вызывающих боль в горле.

```python
import pandas as pd
# Загрузка данных о патогенах
pathogen_data = pd.read_csv('pathogen_data.csv')
# Анализ распространенности патогенов
pathogen_counts = pathogen_data['Pathogen'].value_counts()
```

Сравнение распространенности различных патогенов.

```python
pathogen_counts.plot(kind='bar', title='Распространенность патогенов, вызывающих боль в горле')
```
