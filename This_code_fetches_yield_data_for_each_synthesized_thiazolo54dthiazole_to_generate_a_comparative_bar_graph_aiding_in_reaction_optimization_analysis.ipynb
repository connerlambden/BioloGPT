This notebook demonstrates how to download and visualize yield data from the synthesis of thiazolo[5,4-d]thiazoles across various reaction conditions.

```python
import plotly.graph_objs as go
import plotly.offline as pyo
import pandas as pd

# Data for synthesized thiazolo[5,4-d]thiazole compounds
data = {'Compound': ['TzTz 1', 'TzTz 2', 'TzTz 3', 'TzTz 4', 'TzTz 5', 'TzTz 6', 'TzTz 7', 'TzTz 8'],
        'Yield (%)': [75, 70, 99, 72, 85, 58, 66, 36]}
df = pd.DataFrame(data)

trace = go.Bar(x=df['Compound'], y=df['Yield (%)'], marker=dict(color='#6A0C76'))
layout = go.Layout(title='Yields of Thiazolo[5,4-d]thiazoles', xaxis=dict(title='Compound'), yaxis=dict(title='Yield (%)'))
fig = go.Figure(data=[trace], layout=layout)
pyo.plot(fig, filename='yield_graph.html', auto_open=False)
print('Yield graph generated successfully.')
```
