This notebook will analyze engagement metrics from Instagram posts related to biology, focusing on likes, comments, and shares.

```python
import pandas as pd

# Load the dataset containing Instagram post engagement metrics
data = pd.read_csv('instagram_biology_posts.csv')

# Display the first few rows of the dataset
data.head()
```

The dataset includes columns for post type, likes, comments, and shares. We will analyze the correlation between these metrics.

```python
# Calculate correlation between likes, comments, and shares
correlation = data[['likes', 'comments', 'shares']].corr()
correlation
```

The correlation matrix will help us understand how engagement metrics relate to each other.

```python
# Visualize the correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Engagement Metrics Correlation')
plt.show()
```
