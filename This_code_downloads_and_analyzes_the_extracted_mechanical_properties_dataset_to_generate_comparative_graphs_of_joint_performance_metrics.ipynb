The following notebook section is designed to load the dataset, process the mechanical property values, and create interactive Plotly graphs to compare the performance of different joint configurations.

```python
import pandas as pd
import plotly.express as px

# Load dataset
data = pd.DataFrame([
    {'Configuration': 'Reference', 'Peak Load (kN)': 3.7, 'Energy Absorption Factor': 1},
    {'Configuration': 'Toughened', 'Peak Load (kN)': 5.9, 'Energy Absorption Factor': 2.8},
    {'Configuration': 'Hybrid', 'Peak Load (kN)': 6.5, 'Energy Absorption Factor': 4.5}
])

# Create bar chart
fig = px.bar(data, x='Configuration', y='Peak Load (kN)', color='Configuration',
             title='Comparison of Peak Load Across Configurations',
             labels={'Peak Load (kN)': 'Peak Load (kN)'})
fig.show()

fig2 = px.bar(data, x='Configuration', y='Energy Absorption Factor', color='Configuration',
              title='Energy Absorption Factor Across Configurations',
              labels={'Energy Absorption Factor': 'Absorption Factor'})
fig2.show()
```

The above code provides a quick visual comparative analysis of joint strength and energy absorption between reference, toughened, and hybrid configurations based on the data reported in the study.
