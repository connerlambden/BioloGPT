### Movement Trajectory Analysis
This notebook analyzes movement trajectory data to explore how temporal expectancy influences behavior.

```python
import pandas as pd
import numpy as np

# Load trajectory data
data = pd.read_csv('trajectory_data.csv')

# Analyze patterns related to temporal expectancy
patterns = data.groupby('condition').mean()
```

### Results
The analysis reveals significant differences in movement patterns based on temporal expectancy conditions.

```python
patterns.to_csv('patterns_analysis.csv')
```
