This notebook will analyze RNA-seq data from immune cells of astronauts to identify gene expression changes related to gravity exposure.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load RNA-seq data
rna_data = pd.read_csv('astronauts_rna_seq.csv')

# Analyze gene expression changes
sns.boxplot(x='gravity_condition', y='gene_expression', data=rna_data)
plt.title('Gene Expression Changes Under Different Gravity Conditions')
plt.show()
```

The analysis will help identify specific genes that are upregulated or downregulated in response to lunar and Martian gravity.

```python
# Further analysis to identify significant changes
from scipy import stats

# Perform t-test between gravity conditions
results = stats.ttest_ind(rna_data[rna_data['gravity_condition'] == 'lunar']['gene_expression'], 
rna_data[rna_data['gravity_condition'] == 'martian']['gene_expression'])
print('T-test results:', results)
```
