### Step 1: Data Download and Preprocessing

This step downloads RNA-seq datasets from GEO and applies quality control using Scanpy to filter cells and genes.

```python
import scanpy as sc
adata = sc.read_10x_mtx('path/to/data/')
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)
```

### Step 2: Differential Expression Analysis

Ranking genes based on treatment response using the Wilcoxon rank-sum test to identify metabolic and glycolytic markers.

```python
sc.tl.rank_genes_groups(adata, 'treatment', method='wilcoxon')
sc.pl.rank_genes_groups(adata, n_genes=20, sharey=False)
```

### Step 3: Pathway Enrichment Analysis with GSEApy

This block performs enrichment analysis using KEGG pathways to identify significant metabolic regulators.

```python
import gseapy as gp
ranked_gene_list = adata.uns['rank_genes_groups']['names']['treated']
pre_res = gp.enrichr(gene_list=ranked_gene_list, gene_sets='KEGG_2019_Human', outdir='test/enrichr_kegg')
print(pre_res.results.head())
```
