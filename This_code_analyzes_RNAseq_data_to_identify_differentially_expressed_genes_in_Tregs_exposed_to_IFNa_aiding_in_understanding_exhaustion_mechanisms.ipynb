### RNA-seq Analysis of Tregs Exposed to IFNa
This analysis will focus on identifying genes that are differentially expressed in Tregs treated with IFNa compared to untreated controls.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load RNA-seq data
rna_data = pd.read_csv('Treg_IFNa_RNAseq.csv')

# Perform differential expression analysis
results = stats.ttest_ind(rna_data['IFNa'], rna_data['Control'])

# Identify significant genes
significant_genes = rna_data[results.pvalue < 0.05]
```

### Results Interpretation
The identified significant genes will provide insights into the pathways affected by IFNa treatment in Tregs.

```python
# Display significant genes
print(significant_genes)
```
