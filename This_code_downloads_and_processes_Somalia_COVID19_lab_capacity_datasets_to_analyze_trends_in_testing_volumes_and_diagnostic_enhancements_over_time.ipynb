This notebook retrieves and analyzes data on COVID-19 testing and diagnostic capability improvements in Somalia from provided public datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('somalia_lab_capacity_data.csv')  # Replace with real dataset URL
plt.figure(figsize=(10,6))
plt.plot(data['date'], data['tests_per_day'], marker='o', color='#6A0C76')
plt.title('Somalia COVID-19 Testing Trends')
plt.xlabel('Date')
plt.ylabel('Number of Tests')
plt.grid(True)
plt.show()
```

The code above creates a time-series plot to illustrate the growth in daily COVID-19 tests, reflecting the expansion of laboratory capacity in Somalia.
