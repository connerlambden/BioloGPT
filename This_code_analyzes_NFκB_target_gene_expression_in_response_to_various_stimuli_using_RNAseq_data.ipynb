This notebook will analyze RNA-seq data to identify NF-κB target genes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load RNA-seq data
data = pd.read_csv('nfkb_target_genes.csv')

# Analyze expression levels
sns.boxplot(x='condition', y='expression', data=data)
plt.title('NF-κB Target Gene Expression by Condition')
plt.show()
```

The analysis shows how NF-κB activation affects target gene expression across different conditions.
