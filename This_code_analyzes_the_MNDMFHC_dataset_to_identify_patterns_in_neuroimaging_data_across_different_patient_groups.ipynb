### Data Analysis of MND-MFHC Dataset

```python
import pandas as pd

# Load the dataset
dataset = pd.read_csv('MND_MFHC_data.csv')

# Display basic statistics
print(dataset.describe())
```

### Statistical Analysis
This section will perform statistical tests to compare neuroimaging metrics across groups.

```python
from scipy import stats

# Example: T-test between Facial Paralysis and Healthy Controls
fp_data = dataset[dataset['group'] == 'Facial Paralysis']
hc_data = dataset[dataset['group'] == 'Healthy Control']

t_stat, p_value = stats.ttest_ind(fp_data['metric'], hc_data['metric'])
print(f'T-statistic: {t_stat}, P-value: {p_value}')
```
