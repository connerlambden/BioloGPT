# Analyzing Fluralaner Pharmacokinetics in Chickens
This notebook will explore the pharmacokinetics of fluralaner in chickens and its implications for vector control.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load pharmacokinetic data
pk_data = pd.read_csv('fluralaner_pharmacokinetics.csv')

# Plotting the concentration over time
plt.figure(figsize=(10,6))
plt.plot(pk_data['Time'], pk_data['Concentration'], marker='o')
plt.title('Fluralaner Concentration in Chicken Plasma Over Time')
plt.xlabel('Time (days)')
plt.ylabel('Concentration (ng/mL)')
plt.grid()
plt.show()
```
