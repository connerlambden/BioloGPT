This notebook analyzes woodpecker pecking data to identify trends and correlations.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Bird': ['Bird 1', 'Bird 2', 'Bird 3', 'Bird 4', 'Bird 5', 'Bird 6', 'Bird 7', 'Bird 8'],
        'Sonic_Dissuader_Weight_Removed': [1.1, 42.5, 20.3, 192.5, 69.8, 9.9, 53.8, 47.6],
        'Control_Weight_Removed': [13.8, 21.4, 22.3, 82.9, 25.7, 35.4, 386.5, 219.7]}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Bird'], df['Sonic_Dissuader_Weight_Removed'], label='Sonic Dissuader', alpha=0.6)
plt.bar(df['Bird'], df['Control_Weight_Removed'], label='Control', alpha=0.6)
plt.title('Weight Removed by Woodpeckers')
plt.xlabel('Birds')
plt.ylabel('Weight Removed (g)')
plt.legend()
plt.show()
```

This analysis provides insights into the effectiveness of the Sonic Dissuader compared to control conditions.
