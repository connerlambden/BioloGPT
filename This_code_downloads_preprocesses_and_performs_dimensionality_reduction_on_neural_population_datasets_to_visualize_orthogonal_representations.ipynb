Load and preprocess neural recording data from V4 to explore orthogonal representation features.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Example code to simulate neural population responses
np.random.seed(42)
data = np.random.rand(125, 50)  # simulated responses for 125 images across 50 neurons
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(data)

plt.scatter(reduced_data[:,0], reduced_data[:,1], c='purple', edgecolor='k')
plt.title('PCA of Simulated Neural Responses')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()
```

This notebook performs PCA to identify independent (orthogonal) dimensions within the neural response data, facilitating visualization of feature segregation.
