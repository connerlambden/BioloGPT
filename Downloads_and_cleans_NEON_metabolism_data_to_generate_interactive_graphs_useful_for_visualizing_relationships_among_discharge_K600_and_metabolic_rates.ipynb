Import the NEON metabolism dataset, perform cleaning, and prepare it for statistical visualization.

```python
import pandas as pd
import plotly.express as px

data = pd.read_csv('neon_metabolism_data.csv')
data_clean = data.dropna()
fig = px.scatter(data_clean, x='discharge', y='K600', color='site', title='Discharge vs K600')
fig.show()
```

Visualize site-specific variations in metabolic rates (GPP) across different river systems.

```python
fig = px.box(data_clean, x='site', y='GPP', title='Distribution of GPP per Site')
fig.show()
```
