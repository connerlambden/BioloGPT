We start by loading a dataset containing energy contributions (van der Waals and electrostatic) for various compounds binding to TMPRSS2.

```python
import pandas as pd
import plotly.express as px

# Load the binding energy data (assumed available as 'binding_energies.csv')
df = pd.read_csv('binding_energies.csv')

# Calculate the total binding energy for each compound
 df['total_energy'] = df['vdW_energy'] + df['elec_energy']

# Plot scatter: vdW vs Electrostatic energies
fig = px.scatter(df, x='vdW_energy', y='elec_energy', text='Compound',
                 title='Van der Waals vs Electrostatic Energy Contributions',
                 labels={'vdW_energy':'vdW Energy (kcal/mol)', 'elec_energy':'Electrostatic Energy (kcal/mol)'})
fig.update_traces(textposition='top center')
fig.show()

# Bar plot for total binding energy per compound
fig2 = px.bar(df, x='Compound', y='total_energy', title='Total Binding Energy per Compound')
fig2.show()

# Compute and display correlation between vdW and electrostatic energies
corr = df['vdW_energy'].corr(df['elec_energy'])
print('Correlation between vdW and Electrostatic energies:', corr)
```

This analysis helps quantify how van der Waals and electrostatic interactions correlate, offering insights into the dominant energy contributions governing binding stability.
