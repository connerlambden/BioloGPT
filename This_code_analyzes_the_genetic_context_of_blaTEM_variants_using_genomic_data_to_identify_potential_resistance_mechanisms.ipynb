### Step 1: Import Required Libraries
Import necessary libraries for genomic analysis.

```python
import pandas as pd
import numpy as np
from Bio import SeqIO

# Load genomic data
sequences = SeqIO.parse('genomic_data.fasta', 'fasta')
```

### Step 2: Analyze blaTEM Variants
Identify and analyze the presence of blaTEM variants in the dataset.

```python
blaTEM_variants = []
for seq in sequences:
    if 'blaTEM' in str(seq.seq):
        blaTEM_variants.append(seq.id)

# Convert to DataFrame for analysis
blaTEM_df = pd.DataFrame(blaTEM_variants, columns=['Variant ID'])
```

### Step 3: Visualize Results
Create a visualization of the blaTEM distribution across samples.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(blaTEM_df['Variant ID'], bins=30, color='blue', alpha=0.7)
plt.title('Distribution of blaTEM Variants')
plt.xlabel('Variant ID')
plt.ylabel('Frequency')
plt.show()
```
