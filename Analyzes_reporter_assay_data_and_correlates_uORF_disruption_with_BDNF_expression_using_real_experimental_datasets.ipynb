Download and preprocess raw luciferase reporter data, then perform statistical analysis comparing wild-type and uORF-edited constructs.

```python
import pandas as pd
import plotly.express as px

data = pd.read_csv('bdnf_reporter_data.csv')
fig = px.bar(data, x='Construct', y='Luciferase_Activity', color='Group', title='BDNF Reporter Activity')
fig.show()
```
