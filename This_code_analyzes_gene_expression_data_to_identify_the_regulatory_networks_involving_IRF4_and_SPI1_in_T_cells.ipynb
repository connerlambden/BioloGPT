### Gene Expression Analysis of IRF4 and SPI1 in T Cells
This analysis will utilize RNA-Seq data to explore the expression levels of IRF4 and SPI1 in various T cell subsets.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load gene expression data
data = pd.read_csv('gene_expression_data.csv')

# Filter for IRF4 and SPI1
genes_of_interest = data[data['gene'].isin(['IRF4', 'SPI1'])]

# Plot the expression levels
plt.figure(figsize=(10, 6))
sns.barplot(x='cell_type', y='expression', hue='gene', data=genes_of_interest)
plt.title('Expression Levels of IRF4 and SPI1 in T Cells')
plt.ylabel('Expression Level')
plt.xlabel('T Cell Type')
plt.legend(title='Gene')
plt.show()
```
