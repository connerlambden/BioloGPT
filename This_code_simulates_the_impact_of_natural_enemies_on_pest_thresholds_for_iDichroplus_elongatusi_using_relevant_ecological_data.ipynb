### Simulating Pest Dynamics with Natural Enemies

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
initial_pest_density = 30  # initial density of D. elongatus
natural_enemy_effectiveness = 0.5  # effectiveness of natural enemies

# Simulate pest dynamics
pest_density = initial_pest_density * (1 - natural_enemy_effectiveness)

# Plotting the results
plt.figure(figsize=(10, 5))
plt.plot([0, 10], [initial_pest_density, pest_density], marker='o')
plt.title('Impact of Natural Enemies on Pest Density')
plt.xlabel('Time (days)')
plt.ylabel('Pest Density (individuals/m²)')
plt.grid()
plt.show()
```

This simulation demonstrates how natural enemies can reduce pest densities over time.
