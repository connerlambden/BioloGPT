# mRNA Extraction Kit Analysis
This notebook analyzes the efficiency of various mRNA extraction kits based on yield and purity data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for mRNA extraction kits
data = {
    'Kit': ['MagMAX', 'Dynabeads', 'RNeasy', 'GenElute'],
    'Yield': [90, 70, 85, 75],
    'Purity': [95, 80, 90, 85]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.bar(df['Kit'], df['Yield'], color='blue', alpha=0.7, label='Yield (%)')
plt.bar(df['Kit'], df['Purity'], color='green', alpha=0.5, label='Purity (%)')
plt.title('mRNA Extraction Kit Performance')
plt.xlabel('Kit Name')
plt.ylabel('Percentage (%)')
plt.legend()
plt.show()
```

This analysis provides insights into the performance of different mRNA extraction kits, helping researchers make informed decisions.
