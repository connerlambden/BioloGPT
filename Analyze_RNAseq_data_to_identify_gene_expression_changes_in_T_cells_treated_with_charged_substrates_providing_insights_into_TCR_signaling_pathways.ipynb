This notebook analyzes RNA-seq data from T cells treated with charged substrates to identify differentially expressed genes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load RNA-seq data
data = pd.read_csv('rna_seq_data.csv')

# Analyze differential expression
results = data[data['p_value'] < 0.05]

# Visualize results
plt.figure(figsize=(10,6))
sns.barplot(x='gene', y='log2_fold_change', data=results)
plt.title('Differential Gene Expression in T Cells')
plt.xticks(rotation=90)
plt.show()
```

The results highlight key genes involved in TCR signaling pathways that are affected by charged substrate treatment.

```python
# Save results to CSV
results.to_csv('differential_expression_results.csv', index=False)
```
