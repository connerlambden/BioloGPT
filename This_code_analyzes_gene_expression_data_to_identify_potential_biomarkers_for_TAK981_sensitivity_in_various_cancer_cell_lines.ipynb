### Gene Expression Analysis for TAK-981 Sensitivity

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_data.csv')
sns.boxplot(x='sensitivity', y='gene_expression', data=data)
plt.title('Gene Expression vs TAK-981 Sensitivity')
plt.show()
```

This analysis visualizes the relationship between gene expression levels and sensitivity to TAK-981, helping to identify potential biomarkers.

```python
# Further analysis to identify specific genes associated with sensitivity
sensitive_genes = data[data['sensitivity'] == 'high']['gene'].unique()
print(sensitive_genes)
```
