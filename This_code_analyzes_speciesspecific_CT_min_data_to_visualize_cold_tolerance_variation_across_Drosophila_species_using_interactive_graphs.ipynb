This notebook loads the CT_min dataset extracted from the paper and generates an interactive bar graph to visualize species-specific cold tolerance.

```python
import plotly.graph_objects as go

# Data for species CT_min
species = ['Drosophila montana', 'Drosophila ezoana', 'Drosophila littoralis', 'Drosophila obscura', 'Drosophila virilis', 'Drosophila melanogaster', 'Drosophila rufa', 'Drosophila teissieri', 'Drosophila bunnanda', 'Drosophila bipectinata']
ct_min = [-3.13, -2.39, -1.98, -0.57, 0.8, 2.21, 3.63, 6.67, 7.34, 7.37]

fig = go.Figure([go.Bar(x=species, y=ct_min, marker_color='#6A0C76')])
fig.update_layout(title='CT_min of Drosophila Species', xaxis_title='Species', yaxis_title='CT_min (°C)')
fig.show()
```

The visualization confirms interspecific differences in cold tolerance and supports the review's conclusions on maintaining membrane potential.
