### Análisis de secuencias de genes de <i>Bartonella bacilliformis</i>
Este análisis se centrará en las secuencias de los genes ParC, ParE y RpoB para identificar variaciones que puedan estar asociadas con la resistencia a antibióticos.

```python
import pandas as pd
import numpy as np

# Cargar las secuencias de los genes
sequences = pd.read_csv('sequences.csv')

# Análisis de variaciones en las secuencias
variations = sequences.apply(lambda x: x.str.count('A|T|C|G'))

# Identificar patrones de resistencia
patterns = variations.describe()
```

### Resultados del análisis
Los resultados mostrarán la frecuencia de nucleótidos en las secuencias de los genes analizados, lo que puede ayudar a identificar patrones de resistencia.

```python
# Mostrar los patrones identificados
print(patterns)
```
