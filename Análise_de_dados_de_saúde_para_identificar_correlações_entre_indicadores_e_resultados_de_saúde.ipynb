Análise de dados de saúde para identificar correlações entre indicadores e resultados de saúde.

```python
import pandas as pd

# Carregar dados de saúde
dados = pd.read_csv('dados_saude.csv')

# Análise de correlação
correlacao = dados.corr()
```

Análise de correlação entre indicadores de saúde.

```python
# Visualização da correlação
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(correlacao, annot=True)
plt.title('Correlação entre Indicadores de Saúde')
plt.show()
```
