This notebook details the extraction and visualization of gene expression changes and redox marker levels from the PCB treatment study.

```python
import pandas as pd
import plotly.express as px

data = pd.DataFrame({
    'Gene': ['Gfap', 'Mmp2', 'Tph2', 'Hmox1', 'S100a5'],
    'Fold_Change': [1.5, 1.3, 1.7, 1.4, 0.7]
})
fig = px.bar(data, x='Gene', y='Fold_Change', title='PCB-modulated Gene Expression', color='Gene')
fig.show()
```

The analysis integrates gene expression and oxidative stress marker datasets to provide mechanistic insights into PCB's actions.

```python
import pandas as pd
import plotly.graph_objects as go

markers = pd.DataFrame({
    'Biomarker': ['MDA', 'SOD', 'CAT'],
    'Relative_Level': [1.6, 0.6, 1.0]
})
fig2 = go.Figure(data=[go.Bar(x=markers['Biomarker'], y=markers['Relative_Level'], marker_color='red')])
fig2.update_layout(title='Oxidative Stress Marker Levels', xaxis_title='Biomarker', yaxis_title='Relative Level')
fig2.show()
```
