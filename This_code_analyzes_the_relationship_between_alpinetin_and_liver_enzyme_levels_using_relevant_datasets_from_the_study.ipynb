This notebook analyzes the effects of alpinetin on liver enzyme levels in C57BL/6 mice.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Group': ['Control', 'Ethanol', 'Ethanol + Alpinetin'],
        'ALT': [33.78, 175.0, 59.04],
        'AST': [86.32, 381.1, 193.1],
        'MDA': [0.94, 1.16, 0.89]}
df = pd.DataFrame(data)
df.set_index('Group').plot(kind='bar')
plt.title('Liver Enzyme Levels')
plt.ylabel('Enzyme Activity')
plt.show()
```
