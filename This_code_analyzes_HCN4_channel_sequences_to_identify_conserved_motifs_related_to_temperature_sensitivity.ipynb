### Sequence Analysis of HCN4 Channels
This section will analyze the HCN4 channel sequences to identify conserved motifs that may be responsible for temperature sensitivity.

```python
import pandas as pd
from Bio import SeqIO

# Load HCN4 sequences
sequences = list(SeqIO.parse('hcn4_sequences.fasta', 'fasta'))

# Analyze motifs
motifs = []
for seq in sequences:
    if 'M407' in str(seq.seq) and 'Y409' in str(seq.seq):
        motifs.append(str(seq.id))

# Create DataFrame
motif_df = pd.DataFrame(motifs, columns=['Motif Sequences'])
print(motif_df)
```
