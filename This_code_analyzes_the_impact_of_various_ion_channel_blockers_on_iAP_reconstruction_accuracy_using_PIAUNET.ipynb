### Data Preparation
Load datasets containing eAP/iAP pairs recorded under different ion channel blocker conditions.

```python
import pandas as pd

dataset = pd.read_csv('eap_iap_pairs.csv')
dataset.head()
```

### Model Training
Train PIA-UNET on the prepared dataset to evaluate its performance with different blockers.

```python
from keras.models import load_model

model = load_model('PIA_UNET_model.h5')
# Train the model with the dataset
model.fit(dataset['eAP'], dataset['iAP'], epochs=50)
```
