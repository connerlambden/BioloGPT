This notebook will analyze genetic data from GWAS studies to identify SNPs associated with uveitis and its systemic comorbidities.

```python
import pandas as pd
# Load GWAS summary statistics for uveitis and comorbidities
df_uveitis = pd.read_csv('uveitis_gwas.csv')
df_ibd = pd.read_csv('ibd_gwas.csv')
# Merge datasets on SNPs
df_merged = pd.merge(df_uveitis, df_ibd, on='SNP')
# Analyze associations
results = df_merged[['SNP', 'p_value_uveitis', 'p_value_ibd']].dropna()
```
