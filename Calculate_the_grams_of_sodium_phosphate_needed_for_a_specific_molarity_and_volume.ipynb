This notebook calculates the grams of sodium phosphate required for a given molarity and volume.

```python
# Constants
molarity = 0.050  # mol/L
volume_ml = 250  # mL
molar_mass = 141.96  # g/mol (for Na2HPO4)

# Convert volume to liters
volume_l = volume_ml / 1000  # L

# Calculate mass
mass_g = molarity * volume_l * molar_mass
mass_g
```
