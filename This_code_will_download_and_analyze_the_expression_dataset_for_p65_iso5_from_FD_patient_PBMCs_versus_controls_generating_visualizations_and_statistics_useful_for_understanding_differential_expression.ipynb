This notebook imports the Fabry disease dataset, performs quality control, and produces differential expression plots for the p65 iso5 isoform.

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('fabry_data.csv')  # load dataset
fig = px.box(df, x='Group', y='p65_iso5_expression', title='p65 iso5 Expression in FD Patients vs Controls')
fig.show()
```

The above code generates a box plot comparing p65 iso5 expression between Fabry disease patients and healthy controls, highlighting significant differences.

```python
# Additional analysis
import scipy.stats as stats
patients = df[df['Group']=='FD']['p65_iso5_expression']
controls = df[df['Group']=='Control']['p65_iso5_expression']
t_stat, p_val = stats.ttest_ind(patients, controls)
print('T-statistic:', t_stat, 'P-value:', p_val)
```

This section performs a t-test to statistically compare the expression levels between groups.

```python
# Further visualization
fig2 = px.histogram(df, x='p65_iso5_expression', color='Group', barmode='overlay', title='Distribution of p65 iso5 Expression')
fig2.show()
```
