This notebook will analyze RNA-seq data from Pleurotus ostreatus to identify key genes involved in substrate degradation.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load transcriptome data
transcriptome_data = pd.read_csv('transcriptome_data.csv')

# Analyze gene expression levels
sns.boxplot(x='substrate', y='expression_level', data=transcriptome_data)
plt.title('Gene Expression Levels on Different Substrates')
plt.show()
```

The analysis will help identify which genes are upregulated in response to different substrates, providing insights into metabolic pathways.

```python
# Further analysis of specific genes
specific_genes = transcriptome_data[transcriptome_data['gene'].isin(['gene1', 'gene2'])]
print(specific_genes)
```
