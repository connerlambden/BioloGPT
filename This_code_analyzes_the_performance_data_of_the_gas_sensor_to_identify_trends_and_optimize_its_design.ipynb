### Data Analysis of Gas Sensor Performance

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for gas response
light_intensity = [1, 10, 50, 100, 150]
gas_response = [0.1, 0.5, 0.8, 0.9, 1.0]

# Create DataFrame
data = pd.DataFrame({'Light Intensity (mW/cm²)': light_intensity, 'Gas Response Amplitude': gas_response})

# Plotting
plt.plot(data['Light Intensity (mW/cm²)'], data['Gas Response Amplitude'], marker='o')
plt.title('Gas Response Amplitude vs Light Intensity')
plt.xlabel('Light Intensity (mW/cm²)')
plt.ylabel('Gas Response Amplitude (normalized)')
plt.grid()
plt.show()
```

This analysis helps in understanding how light intensity affects the gas response of the sensor.
