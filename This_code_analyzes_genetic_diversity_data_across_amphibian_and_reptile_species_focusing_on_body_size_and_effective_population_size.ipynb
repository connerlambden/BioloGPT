### Genetic Diversity Analysis of Amphibians and Reptiles
This analysis will explore the relationship between body size and genetic diversity using available datasets.

```python
import pandas as pd

# Load genetic diversity data
data = pd.read_csv('genetic_diversity_data.csv')

# Analyze the correlation between body size and genetic diversity
correlation = data['body_size'].corr(data['genetic_diversity'])
print('Correlation between body size and genetic diversity:', correlation)
```
