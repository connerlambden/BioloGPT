### Cytotoxicity Analysis of Sphingomyelin Nanoemulsions and Resveratrol

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for IC50 values
data = {'Sample': ['SNs (A549)', 'SNs (HEK293)', 'RSV (free)', 'RSV (nanoencapsulated)'],
        'IC50 (µg/mL)': [0.89, 1.133, 118.54, 85.74]}
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['Sample'], df['IC50 (µg/mL)'], color=['blue', 'orange', 'green', 'red'])
plt.title('IC50 Values for SNs and RSV')
plt.xlabel('Samples')
plt.ylabel('IC50 (µg/mL)')
plt.show()
```

### Discussion
This analysis provides a visual representation of the cytotoxicity of SNs and RSV, highlighting their potential as effective drug delivery systems.
