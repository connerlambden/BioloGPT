### Genetic Diversity Analysis of Visayan Spotted Deer
This notebook analyzes genetic diversity data to understand the impact of habitat fragmentation.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('genetic_diversity_data.csv')
plt.figure(figsize=(10, 6))
plt.bar(data['Population'], data['Genetic_Diversity'])
plt.title('Genetic Diversity in Visayan Spotted Deer')
plt.xlabel('Population')
plt.ylabel('Genetic Diversity')
plt.show()
```
