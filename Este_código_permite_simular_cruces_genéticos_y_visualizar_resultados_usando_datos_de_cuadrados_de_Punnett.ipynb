Simulación de cruces genéticos usando cuadrados de Punnett.

```python
import numpy as np
import pandas as pd

# Definir genotipos de los padres
padre1 = ['A', 'A']
padre2 = ['a', 'a']

# Generar combinaciones posibles
combinaciones = [f'{p1}{p2}' for p1 in padre1 for p2 in padre2]

# Crear DataFrame para visualizar resultados
resultados = pd.DataFrame(combinaciones, columns=['Genotipo'])
resultados['Fenotipo'] = resultados['Genotipo'].apply(lambda x: 'Dominante' if 'A' in x else 'Recesivo')
resultados
```

Visualización de resultados de la simulación.

```python
import matplotlib.pyplot as plt

# Contar fenotipos
conteo = resultados['Fenotipo'].value_counts()

# Gráfico de barras
plt.bar(conteo.index, conteo.values)
plt.title('Distribución de Fenotipos')
plt.xlabel('Fenotipo')
plt.ylabel('Cantidad')
plt.show()
```
