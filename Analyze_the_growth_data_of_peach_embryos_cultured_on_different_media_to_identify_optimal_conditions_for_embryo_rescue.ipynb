This notebook will analyze the growth data of peach embryos cultured on various media, focusing on germination rates and plant development.

```python
import pandas as pd

data = {
    'Genotype': ['Mayfire Nectarine', 'P51-4 Peach', 'K242-45 Plum'],
    'Initial Length (mm)': [5.1, 6.1, 6.0],
    'Final Length (mm)': [7.0, 7.7, 7.4],
    'Germination Rate (%)': [98, 100, 98]
}

df = pd.DataFrame(data)

# Calculate percent change in length

df['Percent Change Length'] = ((df['Final Length (mm)'] - df['Initial Length (mm)']) / df['Initial Length (mm)']) * 100

df
```

The above code calculates the percent change in length for each genotype, providing insights into the effectiveness of the culture media.

```python
# Display the DataFrame
print(df)
```
