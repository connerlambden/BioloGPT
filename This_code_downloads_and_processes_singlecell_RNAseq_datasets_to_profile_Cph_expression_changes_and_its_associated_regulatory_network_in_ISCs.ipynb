Extract and analyze single-cell RNA-seq expression data for Cph in Drosophila ISCs.

```python
import scanpy as sc
adata = sc.read('path_to_data.h5ad')
data = adata[adata.obs['cell_type'] == 'ISC']
sc.pp.normalize_total(data, target_sum=1e4)
sc.pp.log1p(data)
sc.tl.pca(data)
sc.pl.pca(data, color=['Cph'])
```
