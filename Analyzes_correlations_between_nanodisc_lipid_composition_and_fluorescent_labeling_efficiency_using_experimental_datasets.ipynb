This notebook imports and merges lipid composition data with labeling efficiency metrics to calculate Pearson correlation coefficients.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Import datasets
lipid_data = pd.read_csv('lipid_composition.csv')
labeling_data = pd.read_csv('labeling_efficiency.csv')

# Merge datasets on protein_id
merged = pd.merge(lipid_data, labeling_data, on='protein_id')

# Compute Pearson correlation
corr, p_val = pearsonr(merged['lipid_fraction'], merged['labeling_efficiency'])
print('Correlation:', corr, 'P-value:', p_val)

# Visualize the relationship
plt.figure(figsize=(8,6))
plt.scatter(merged['lipid_fraction'], merged['labeling_efficiency'], color='#6A0C76')
plt.title('Lipid Fraction vs. Labeling Efficiency')
plt.xlabel('Lipid Fraction')
plt.ylabel('Labeling Efficiency')
plt.grid(True)
plt.show()
```

The resulting analysis provides a quantitative measure of how variations in lipid composition affect labeling efficiency, guiding further experimental design.
