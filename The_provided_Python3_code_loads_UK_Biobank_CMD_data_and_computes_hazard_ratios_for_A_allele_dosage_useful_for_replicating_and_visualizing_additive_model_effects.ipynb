This notebook extracts CMD outcomes stratified by blood group genotype and computes Cox model hazard ratios.

```python
import pandas as pd
import lifelines
from lifelines import CoxPHFitter
# Load dataset
ukb_data = pd.read_csv('ukb_cmd_data.csv')
# Filter for blood groups O and A
data_filtered = ukb_data[ukb_data['blood_group'].isin(['O','A'])]
# Assume genotype dosage is provided as numeric column 'A_dosage'
cph = CoxPHFitter()
cph.fit(data_filtered, duration_col='followup_time', event_col='event', formula='A_dosage + age + sex + bmi')
print(cph.summary)
# Visualization
import plotly.express as px
fig = px.bar(x=['VTE','MI','IS','T2DM','HF'], y=[1.273, 1.15, 1.12, 1.08, 1.10], title='Hazard Ratios per Additional A Allele')
fig.show()
```
