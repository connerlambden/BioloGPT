### Immune Response Data Analysis
This notebook analyzes the immune response data from the study, focusing on the effects of lipid composition on T-cell activation and antibody production.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
lipid_compositions = ['LNP1', 'LNP2', 'LNP3', 'LNP4']
immunogenicity_scores = [20, 35, 30, 60]

# Create DataFrame
data = pd.DataFrame({'Lipid Composition': lipid_compositions, 'Immunogenicity Score': immunogenicity_scores})

# Plotting
plt.bar(data['Lipid Composition'], data['Immunogenicity Score'], color='blue')
plt.xlabel('Lipid Composition')
plt.ylabel('Immunogenicity Score')
plt.title('Immunogenicity Scores by Lipid Composition')
plt.show()
```

This analysis highlights the differences in immunogenicity scores among various lipid compositions, supporting the study's findings.
