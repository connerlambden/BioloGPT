This notebook loads structural and kinetic data for Kluyveromyces lactis β-galactosidase mutants to analyze the effects of domain insertions.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example: Load mutation impact data
data = pd.read_csv('enzyme_mutation_data.csv')
plt.figure(figsize=(10,6))
plt.plot(data['Mutation'], data['Km'], marker='o', linestyle='-', color='#6A0C76')
plt.xlabel('Mutation', fontsize=14)
plt.ylabel('Km (mM)', fontsize=14)
plt.title('Impact of Domain Insertions on Km', fontsize=16)
plt.grid(True)
plt.show()
```
