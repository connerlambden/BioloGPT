### Patient Data Analysis for MRONI Risk Factors

```python
import pandas as pd

# Load patient data
patient_data = pd.read_csv('patient_data.csv')

# Analyze risk factors for MRONI
risk_analysis = patient_data.groupby('medication').agg({'MRONI': 'mean'})
print(risk_analysis)
```
