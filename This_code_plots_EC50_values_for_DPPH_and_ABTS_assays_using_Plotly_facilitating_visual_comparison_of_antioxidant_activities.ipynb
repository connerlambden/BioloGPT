This section outlines steps to download relevant datasets and generate a comparative bar chart for EC50 data.

```python
import plotly.graph_objects as go

complexes = ['Complex 1', 'Complex 2', 'Complex 3']
dpph = [1.14, 1.90, 3.56]
abts = [5.78, 7.01, 11.12]

fig = go.Figure(data=[
    go.Bar(name='DPPH EC50', x=complexes, y=dpph, marker_color='#6A0C76'),
    go.Bar(name='ABTS EC50', x=complexes, y=abts, marker_color='#A83279')
])

fig.update_layout(barmode='group', title='EC50 Values Comparison', xaxis_title='Complexes', yaxis_title='EC50 (µM)')
fig.show()
```
