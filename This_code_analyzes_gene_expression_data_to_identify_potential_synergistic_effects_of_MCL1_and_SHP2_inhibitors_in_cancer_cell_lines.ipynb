### Step 1: Load necessary libraries and datasets
Import libraries for data analysis and visualization. Load gene expression datasets for cancer cell lines treated with MCL-1 and SHP2 inhibitors.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_data.csv')
sns.boxplot(x='treatment', y='gene_expression', data=data)
plt.title('Gene Expression Levels by Treatment')
plt.show()
```
