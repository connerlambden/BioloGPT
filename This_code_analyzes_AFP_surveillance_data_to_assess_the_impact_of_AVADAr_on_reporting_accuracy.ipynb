### Data Analysis of AFP Surveillance
This section will analyze the AFP surveillance data to evaluate the effectiveness of AVADAr.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = {'Year': [2016, 2017, 2018, 2019, 2020, 2021],
        'Total Cases': [364, 682, 973, 905, 585, 625],
        'cVDPV2 Cases': [0, 0, 9, 1, 10, 13]}

df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(df['Year'], df['Total Cases'], label='Total Cases', color='blue')
plt.bar(df['Year'], df['cVDPV2 Cases'], label='cVDPV2 Cases', color='orange')
plt.title('AFP Cases Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.legend()
plt.show()
```
