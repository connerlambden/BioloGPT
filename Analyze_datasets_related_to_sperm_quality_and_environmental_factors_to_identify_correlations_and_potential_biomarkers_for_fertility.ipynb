This notebook will analyze sperm quality data in relation to environmental exposure datasets.

```python
import pandas as pd
# Load datasets
sperm_data = pd.read_csv('sperm_quality.csv')
environmental_data = pd.read_csv('environmental_exposure.csv')
# Merge datasets for analysis
merged_data = pd.merge(sperm_data, environmental_data, on='sample_id')
# Perform correlation analysis
correlation = merged_data.corr()
```

The correlation matrix will help identify significant relationships between environmental factors and sperm quality.

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Visualize correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Environmental Factors and Sperm Quality')
plt.show()
```
