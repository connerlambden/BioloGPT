This notebook will analyze the comedogenicity scores of various ingredients and their reported skin reactions.

```python
import pandas as pd

# Load comedogenicity data
data = {'Ingredient': ['PPG-2 myristyl propionate', 'D & C Red No. 33', 'sulfated castor oil', 'isopropyl myristate', 'precipitated sulfur'],
        'Comedogenicity Grade': [5, 3, 5, 5, 2],
        'Irritation': ['marked surface irritation', 'significant irritation', 'significant irritation', 'significant irritation', 'minimal irritation']}

df = pd.DataFrame(data)

# Analyze the data
print(df.describe() )
```

The analysis provides insights into which ingredients are most likely to cause skin irritation based on their comedogenicity scores.

```python
# Visualizing the data
import matplotlib.pyplot as plt

plt.bar(df['Ingredient'], df['Comedogenicity Grade'], color='skyblue')
plt.xlabel('Ingredients')
plt.ylabel('Comedogenicity Grade')
plt.title('Comedogenicity of Skincare Ingredients')
plt.xticks(rotation=45)
plt.show()
```
