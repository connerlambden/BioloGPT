This notebook will analyze DNA sequences to identify regions that form double-stranded structures and assess their stability based on GC content.

```python
import pandas as pd

def analyze_dna_sequences(sequences):
    results = []
    for seq in sequences:
        gc_content = (seq.count('G') + seq.count('C')) / len(seq)
        results.append({'sequence': seq, 'gc_content': gc_content})
    return pd.DataFrame(results)

# Example usage
sequences = ['ATCGATCG', 'GCGCGC', 'ATATATAT']
results_df = analyze_dna_sequences(sequences)
results_df
```

The results will show the GC content of each DNA sequence, which is a critical factor in determining the stability of double-stranded DNA.

```python
results_df.to_csv('dna_analysis_results.csv', index=False)
```
