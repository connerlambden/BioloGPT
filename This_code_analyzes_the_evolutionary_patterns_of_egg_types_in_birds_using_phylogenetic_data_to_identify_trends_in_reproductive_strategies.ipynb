### Analyzing Egg Type Evolution in Birds
This notebook will explore the evolutionary patterns of soft and hard-shelled eggs in birds using phylogenetic data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load phylogenetic data
phylo_data = pd.read_csv('bird_phylogeny.csv')

# Analyze egg type distribution
egg_types = phylo_data['egg_type'].value_counts()

# Plot the distribution of egg types
plt.figure(figsize=(10, 6))
plt.bar(egg_types.index, egg_types.values, color=['blue', 'orange'])
plt.title('Distribution of Egg Types in Birds')
plt.xlabel('Egg Type')
plt.ylabel('Number of Species')
plt.show()
```
