Analyse des données de méiose pour identifier les facteurs influençant le pachyte et le crossing over.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Échantillon': ['Homme', 'Femme'], 'Foyers': [50.9, 95]}
df = pd.DataFrame(data)
df.plot(kind='bar', x='Échantillon', y='Foyers', color='#6A0C76')
plt.title('Distribution des foyers dans les spermatocytes pachytene')
plt.ylabel('Nombre de foyers')
plt.show()
```
