This notebook imports the dataset of sucrose and cytokinin treatments, plotting the percentage of plant induction vs sucrose concentration.

```python
import pandas as pd
import plotly.express as px
df = pd.DataFrame([
    {'Sucrose':'20 g/L', 'Induction':10, 'Microtubers':1.03, 'Mass':34.0, 'Diameter':2.5},
    {'Sucrose':'80 g/L', 'Induction':63.3, 'Microtubers':1.9, 'Mass':75, 'Diameter':5.3},
    {'Sucrose':'120 g/L', 'Induction':46.6, 'Microtubers':1.13, 'Mass':25.7, 'Diameter':2.6}
])
fig = px.bar(df, x='Sucrose', y='Induction', title='Percentage of Plant Induction vs Sucrose Concentration', labels={'Induction':'Induction (%)'})
fig.show()
```
