### Analyzing Home-Range Sizes and Carrying Capacity
This notebook will simulate the effects of varying home-range sizes on species populations in fragmented landscapes.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate carrying capacity based on home-range size
home_range_sizes = np.array([1, 2, 3, 4, 5])  # Example sizes
carrying_capacity = 100 - (home_range_sizes ** 2)  # Hypothetical relationship

plt.plot(home_range_sizes, carrying_capacity, marker='o')
plt.title('Carrying Capacity vs Home-Range Size')
plt.xlabel('Home-Range Size')
plt.ylabel('Carrying Capacity')
plt.grid(True)
plt.show()
```

### Discussion
This simulation illustrates how increasing home-range sizes can lead to a decrease in carrying capacity, reflecting the challenges faced by larger species in fragmented landscapes.
