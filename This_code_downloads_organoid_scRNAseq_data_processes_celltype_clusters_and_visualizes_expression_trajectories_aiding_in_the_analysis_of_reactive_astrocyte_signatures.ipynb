Download and preprocess the scRNA-seq dataset from the specified repository to analyze differential expression.

```python
import scanpy as sc
adata = sc.read_h5ad('path/to/organoid_data.h5ad')
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)
sc.tl.pca(adata, svd_solver='arpack')
sc.pl.pca(adata, color='cell_type')
```
