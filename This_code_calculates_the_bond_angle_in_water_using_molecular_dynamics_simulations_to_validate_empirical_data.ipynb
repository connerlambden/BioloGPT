# Calculate the bond angle in water using molecular dynamics simulations.

```python
# Import necessary libraries
import numpy as np

# Function to calculate bond angle
def calculate_bond_angle(o, h1, h2):
    # Vector from O to H1
    vec1 = h1 - o
    # Vector from O to H2
    vec2 = h2 - o
    # Calculate angle using dot product
    cos_angle = np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
    angle = np.arccos(cos_angle) * (180 / np.pi)  # Convert to degrees
    return angle

# Example coordinates for water molecule
O = np.array([0, 0, 0])
H1 = np.array([0.96, 0, 0])
H2 = np.array([0.96 * np.cos(104.5 * np.pi / 180), 0.96 * np.sin(104.5 * np.pi / 180), 0])

# Calculate bond angle
angle = calculate_bond_angle(O, H1, H2)
angle
```

# The calculated bond angle should be approximately 104.5 degrees.

```python
print(f'The bond angle in water is approximately: {angle:.2f} degrees')
```
