Analyse des données de recherche en neurosciences pour identifier des patterns et des corrélations entre les spécialités.

```python
import pandas as pd

# Charger les données de recherche
research_data = pd.read_csv('neuroscience_research.csv')

# Analyser les corrélations entre les spécialités
correlations = research_data.corr()
```

Cette analyse permet d'identifier les relations entre différentes spécialités en neurosciences.

```python
# Visualiser les corrélations
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(correlations, annot=True)
plt.title('Corrélations entre spécialités en neurosciences')
plt.show()
```
