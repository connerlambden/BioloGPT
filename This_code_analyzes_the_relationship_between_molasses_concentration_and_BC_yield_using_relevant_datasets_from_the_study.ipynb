# Analysis of BC Yield vs. Molasses Concentration
This notebook will analyze the impact of varying molasses concentrations on the yield of bacterial cellulose.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from the study
data = {
    'Molasses Concentration (g/L)': [50, 70, 90, 110],
    'BC Yield (g/L)': [2.68, 3.15, 3.88, 3.95]
}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Molasses Concentration (g/L)'], df['BC Yield (g/L)'], marker='o')
plt.title('BC Yield vs. Molasses Concentration')
plt.xlabel('Molasses Concentration (g/L)')
plt.ylabel('BC Yield (g/L)')
plt.grid()
plt.show()
```

This analysis provides insights into how varying molasses concentrations affect the yield of bacterial cellulose.
