This notebook analyzes extracted data from the gelatin-based Gum Arabic film study and biodegradation efficiency data from wheat bran experiments.

```python
import pandas as pd
import plotly.express as px

data = pd.read_json('your_data_source.json')
fig = px.bar(data, x='pathogen', y='inhibition_zone_mm', color='effluent_type', barmode='group', title='Antimicrobial Efficacy of GEL/GA Films')
fig.show()
```

Next, we analyze biodegradation efficiency data across different substrates ensuring reproducibility and statistical significance.

```python
import plotly.graph_objects as go

data_bd = pd.read_json('your_biodegradation_data.json')
fig2 = go.Figure()
for substrate in data_bd['Substrate ID'].unique():
    df_sub = data_bd[data_bd['Substrate ID'] == substrate]
    fig2.add_trace(go.Scatter(x=df_sub['Bran Content (%)'], y=df_sub['Biodegradation Efficiency (%)'], mode='lines+markers', name=substrate))
fig2.update_layout(title='Biodegradation Efficiency Across Substrates', xaxis_title='Bran Content (%)', yaxis_title='Efficiency (%)')
fig2.show()
```
