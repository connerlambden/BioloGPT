This notebook will analyze data on bacterial strains found in scalp and eye infections, focusing on their prevalence and potential transmission routes.

```python
import pandas as pd

# Load datasets on bacterial strains from scalp and eye infections
dataset_scalp = pd.read_csv('scalp_infections.csv')
dataset_eye = pd.read_csv('eye_infections.csv')

# Analyze common strains
common_strains = pd.merge(dataset_scalp, dataset_eye, on='strain', how='inner')
common_strains.to_csv('common_bacterial_strains.csv', index=False) # Save results
```

The analysis identifies common bacterial strains between scalp and eye infections, providing insights into potential transmission pathways.
