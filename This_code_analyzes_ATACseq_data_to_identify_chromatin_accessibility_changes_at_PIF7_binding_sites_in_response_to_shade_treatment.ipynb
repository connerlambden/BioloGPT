This notebook will analyze ATAC-seq data to assess chromatin accessibility at PIF7 binding sites before and after shade treatment.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load ATAC-seq data
atac_data = pd.read_csv('atac_seq_data.csv')

# Analyze accessibility changes
plt.figure(figsize=(10,6))
plt.plot(atac_data['Time'], atac_data['Accessibility'], marker='o')
plt.title('Chromatin Accessibility at PIF7 Binding Sites')
plt.xlabel('Time (h)')
plt.ylabel('Accessibility')
plt.grid()
plt.show()
```

The plot above illustrates the changes in chromatin accessibility at PIF7 binding sites over time, highlighting the impact of shade treatment.
