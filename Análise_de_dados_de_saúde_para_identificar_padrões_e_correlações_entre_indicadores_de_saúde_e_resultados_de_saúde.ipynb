Análise de dados de saúde para identificar padrões e correlações entre indicadores de saúde e resultados de saúde.

```python
import pandas as pd

data = pd.read_csv('health_data.csv')

# Análise de correlação
correlation = data.corr()
```

A análise de correlação ajuda a entender como diferentes indicadores de saúde estão inter-relacionados.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(correlation, annot=True)
plt.title('Correlação entre Indicadores de Saúde')
plt.show()
```
