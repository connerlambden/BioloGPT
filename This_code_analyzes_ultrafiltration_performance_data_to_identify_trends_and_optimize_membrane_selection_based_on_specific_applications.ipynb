### Ultrafiltration Performance Analysis
This notebook analyzes the performance of various ultrafiltration membranes based on experimental data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for ultrafiltration performance
data = {'Experiment': ['BMX5-1', 'BMX5-2', 'BMX5-3', 'BMX5-4', 'BMX5-5'],
        'Q_F (L/h)': [1.25, 1.25, 1.25, 1.25, 2.4],
        'TMP (bar)': [0.15, 0.4, 0.6, 0.8, 0.4]}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 5))
plt.bar(df['Experiment'], df['Q_F (L/h)'], label='Q_F (L/h)', alpha=0.6)
plt.bar(df['Experiment'], df['TMP (bar)'], label='TMP (bar)', alpha=0.6)
plt.title('Ultrafiltration Performance Data')
plt.xlabel('Experiments')
plt.ylabel('Values')
plt.legend()
plt.show()
```

This analysis provides insights into the performance of different ultrafiltration membranes, aiding in the selection process for specific applications.
