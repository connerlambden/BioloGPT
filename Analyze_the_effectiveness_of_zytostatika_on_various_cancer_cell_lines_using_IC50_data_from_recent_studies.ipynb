This notebook will analyze the IC50 values of different zytostatika on various cancer cell lines to determine their relative effectiveness.

```python
import pandas as pd

# Sample data for IC50 values
data = {
    'Cell Line': ['NALM-6', 'MOLT-3', 'HL-60', 'DAUDI'],
    'IC50 (µmol/l)': [4.0, 4.3, 0.2, 0.8]
}

# Create DataFrame
df = pd.DataFrame(data)

# Analyze effectiveness
print(df.describe() )
```

The analysis provides insights into which zytostatika are most effective against specific cancer cell lines based on their IC50 values.

```python
# Visualizing the data
import matplotlib.pyplot as plt

plt.bar(df['Cell Line'], df['IC50 (µmol/l)'], color='purple')
plt.title('IC50 Values of Zytostatika on Cancer Cell Lines')
plt.xlabel('Cell Line')
plt.ylabel('IC50 (µmol/l)')
plt.show()
```
