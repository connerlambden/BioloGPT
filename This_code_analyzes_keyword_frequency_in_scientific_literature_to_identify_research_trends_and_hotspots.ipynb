### Keyword Frequency Analysis
This notebook analyzes the frequency of keywords in selected scientific papers to identify trends and research hotspots.

```python
import pandas as pd

# Load dataset containing keywords
keywords_df = pd.read_csv('keywords_data.csv')

# Analyze frequency of each keyword
keyword_counts = keywords_df['keyword'].value_counts()

# Display the top 10 keywords
print(keyword_counts.head(10))
```

### Results
The analysis reveals the most frequently used keywords, indicating current research trends.

```python
# Plotting the keyword frequency
import matplotlib.pyplot as plt

keyword_counts.head(10).plot(kind='bar')
plt.title('Top 10 Keywords in Scientific Literature')
plt.xlabel('Keywords')
plt.ylabel('Frequency')
plt.show()
```
