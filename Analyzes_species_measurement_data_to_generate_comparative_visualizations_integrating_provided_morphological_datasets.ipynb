This notebook downloads the species measurement data, cleans it, and generates comparative bar plots for body length and width.

```python
import plotly.graph_objs as go
import plotly.offline as pyo

# Data from the paper
species = ['tajikistanensis', 'asetosus']
body_length = [450, 682]
body_width = [277, 420]

trace1 = go.Bar(x=species, y=body_length, name='Body Length')
trace2 = go.Bar(x=species, y=body_width, name='Body Width')
layout = go.Layout(barmode='group', title='Holotype Measurements Comparison')
fig = go.Figure(data=[trace1, trace2], layout=layout)
pyo.plot(fig, filename='species_measurements.html')
```
