### User Feedback Data Analysis
This notebook analyzes user feedback data collected from prosthetic hand users to evaluate the impact of multimodal feedback on embodiment perception.

```python
import pandas as pd

# Load user feedback data
data = pd.read_csv('user_feedback.csv')

# Analyze correlations between feedback types and embodiment scores
correlation = data.corr()
```

### Correlation Analysis
The correlation matrix will help identify which types of feedback are most strongly associated with improved embodiment perception.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Feedback Types and Embodiment Scores')
plt.show()
```
