This notebook analyzes the performance of various yeast strains based on glucose consumption and ethanol production.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Strain': ['ACY8', 'ACY19', 'ACY34', 'ACY84'],
        'Ethanol Production (g/L/h)': [0.120, 0.142, 0.263, 0.233],
        'Glucose Consumption (g/L/h)': [1.133, 1.215, 1.604, 1.115]}
df = pd.DataFrame(data)

plt.bar(df['Strain'], df['Ethanol Production (g/L/h)'], color='b', label='Ethanol Production')
plt.bar(df['Strain'], df['Glucose Consumption (g/L/h)'], color='r', label='Glucose Consumption', alpha=0.5)
plt.title('Yeast Strain Performance')
plt.xlabel('Yeast Strain')
plt.ylabel('Rate (g/L/h)')
plt.legend()
plt.show()
```

This analysis provides insights into which yeast strains are most effective for fermentation based on their metabolic performance.
