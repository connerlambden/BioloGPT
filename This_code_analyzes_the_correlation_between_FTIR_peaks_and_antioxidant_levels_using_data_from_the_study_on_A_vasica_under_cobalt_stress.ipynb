# Analysis of FTIR Peaks and Antioxidant Responses in A. vasica
This notebook will analyze the correlation between FTIR peaks and antioxidant responses in A. vasica under cobalt stress.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
cobalt_data = pd.read_csv('cobalt_study_data.csv')

# Analyze the correlation
correlation = cobalt_data.corr()
print(correlation)

# Visualize the results
plt.figure(figsize=(10, 6))
plt.title('Correlation between FTIR Peaks and Antioxidant Responses')
plt.imshow(correlation, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.show()
```
