### Analyze STAT3 and FOXP3 Expression in Tregs

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load single-cell RNA-seq data
sc_data = pd.read_csv('path_to_data.csv')

# Filter for Tregs
treg_data = sc_data[sc_data['cell_type'] == 'Treg']

# Plot expression levels
plt.figure(figsize=(10,6))
sns.boxplot(data=treg_data[['STAT3', 'FOXP3']])
plt.title('Expression Levels of STAT3 and FOXP3 in Tregs')
plt.ylabel('Expression Level')
plt.show()
```
