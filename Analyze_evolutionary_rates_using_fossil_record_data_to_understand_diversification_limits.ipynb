This notebook will analyze extinction and origination rates from fossil records to explore evolutionary speed limits.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for extinction and origination rates
rates_data = {
    'Type': ['Extinction (families)', 'Origination (families)', 'Extinction (genera)', 'Origination (genera)'],
    'Rate': [25.1, 34.8, 149, 197]
}

# Create DataFrame
rates_df = pd.DataFrame(rates_data)

# Plotting
plt.bar(rates_df['Type'], rates_df['Rate'], color=['red', 'green', 'blue', 'orange'])
plt.title('Extinction and Origination Rates')
plt.ylabel('Rates per Interval')
plt.show()
```

This analysis provides insights into how extinction and origination rates influence evolutionary dynamics.
