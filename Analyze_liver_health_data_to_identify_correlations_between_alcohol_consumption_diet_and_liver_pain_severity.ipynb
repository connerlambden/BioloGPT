This notebook analyzes liver health data to explore the relationship between alcohol consumption, dietary habits, and reported liver pain.

```python
import pandas as pd
# Load liver health dataset
dataset = pd.read_csv('liver_health_data.csv')
# Analyze correlations
correlation = dataset.corr()
```

The correlation matrix provides insights into how different factors relate to liver pain severity.

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Visualize correlations
sns.heatmap(correlation, annot=True)
plt.title('Correlation Matrix of Liver Health Factors')
plt.show()
```
