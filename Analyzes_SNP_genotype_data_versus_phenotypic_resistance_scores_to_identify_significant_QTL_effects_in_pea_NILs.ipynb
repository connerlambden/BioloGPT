This notebook loads SNP genotype and phenotype datasets to perform linear mixed model analysis, correlating specific QTL markers with observed field resistance.

```python
import pandas as pd
import statsmodels.api as sm
# Load SNP genotype data and field phenotype data
geno = pd.read_csv('geno_data.csv')
pheno = pd.read_csv('pheno_data.csv')
# Merge datasets on LineID
data = pd.merge(geno, pheno, on='LineID')
# Fit a mixed linear model considering genetic background as a random effect
model = sm.MixedLM(data['DiseaseScore'], data[['Ae-Ps7.6', 'Ae-Ps4.1', 'Ae-Ps5.1']], groups=data['GeneticBackground'])
result = model.fit()
print(result.summary())
```

The output offers insights into the significance and effect sizes of combined QTL on disease severity, guiding breeding decisions.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(data['Ae-Ps7.6'], data['DiseaseScore'], color='#6A0C76')
plt.xlabel('Marker Score: Ae-Ps7.6')
plt.ylabel('Disease Severity')
plt.title('Correlation between Ae-Ps7.6 and Field Resistance')
plt.show()
```
