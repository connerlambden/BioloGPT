### Data Collection and Preparation
- Gather morphological measurements of intertegular distance and proboscis length from tropical bee specimens.

```python
import pandas as pd

# Load the dataset
url = 'https://github.com/kilian-fru/tropTongue-analysis'
data = pd.read_csv(url + '/morphological_data.csv')

# Display the first few rows of the dataset
data.head()
```

### Model Development
- Develop allometric models using linear regression to predict proboscis length based on intertegular distance.

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare the data
X = data[['intertegular_distance']]
y = data['proboscis_length']

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Predict proboscis length
predictions = model.predict(X)
```
