### Genetic Analysis of Color Variation in Reptiles

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load genetic data related to Mc1r gene
mc1r_data = pd.read_csv('mc1r_genetic_data.csv')

# Analyze polymorphisms and their association with color morphs
sns.boxplot(x='color_morph', y='allele_frequency', data=mc1r_data)
plt.title('Allele Frequency by Color Morph')
plt.show()
```

This analysis will help identify the genetic basis of color variation in reptiles and its correlation with substrate color.

```python
# Further analysis can include statistical tests to assess significance
from scipy import stats
blanched = mc1r_data[mc1r_data['color_morph'] == 'blanched']['allele_frequency']
wild_type = mc1r_data[mc1r_data['color_morph'] == 'wild_type']['allele_frequency']

# Perform t-test
t_stat, p_value = stats.ttest_ind(blanched, wild_type)
print(f'T-statistic: {t_stat}, P-value: {p_value}')
```
