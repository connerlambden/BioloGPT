### Immune Response Data Analysis
This notebook will analyze the immunogenicity data from the SAFER cohort and other relevant studies.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
vaccine_data = {'Time Point': ['T1', 'T2', 'T3', 'T4'], 'IgG Titer': [6.26, 73.01, 200.0, 904.92]}
df = pd.DataFrame(vaccine_data)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.bar(df['Time Point'], df['IgG Titer'], color='blue')
plt.title('IgG Titer Response Over Time')
plt.xlabel('Time Points')
plt.ylabel('IgG Titer (BAU/mL)')
plt.show()
```

### Discussion
This analysis provides insights into the immune response over time following vaccination, highlighting the importance of monitoring IgG titers.
