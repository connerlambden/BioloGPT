This notebook analyzes blood pressure data from various mammalian species.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Species': ['Giraffe', 'Elephant', 'Human', 'Dog'], 'Systolic_BP': [208, 120, 120, 80], 'Diastolic_BP': [100, 60, 80, 50]}
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Species'], df['Systolic_BP'], color='b', label='Systolic BP')
plt.bar(df['Species'], df['Diastolic_BP'], color='r', label='Diastolic BP', alpha=0.5)
plt.title('Blood Pressure in Various Mammals')
plt.xlabel('Species')
plt.ylabel('Blood Pressure (mm Hg)')
plt.legend()
plt.show()
```

This analysis provides insights into the differences in blood pressure across species.
