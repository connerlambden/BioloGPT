This notebook will analyze nutrient uptake data from glacier ice algae, focusing on the effects of phosphorus and nitrogen on growth rates.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load nutrient uptake data
data = pd.read_csv('nutrient_uptake.csv')

# Analyze the data
plt.figure(figsize=(10, 6))
plt.scatter(data['phosphorus'], data['growth_rate'], color='blue')
plt.title('Nutrient Uptake vs Growth Rate')
plt.xlabel('Phosphorus Concentration (µM)')
plt.ylabel('Growth Rate (cells/mL/day)')
plt.show()
```

This analysis will help identify the optimal nutrient conditions for glacier ice algae growth.
