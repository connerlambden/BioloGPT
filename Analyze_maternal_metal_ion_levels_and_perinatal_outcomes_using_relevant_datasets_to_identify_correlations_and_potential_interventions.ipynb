This notebook will analyze the relationship between maternal metal ion levels and perinatal outcomes using data from the OBESO cohort.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
url = 'https://example.com/obeso_data.csv'
data = pd.read_csv(url)

# Analyze correlations
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation between Metal Ion Levels and Perinatal Outcomes')
plt.show()
```
