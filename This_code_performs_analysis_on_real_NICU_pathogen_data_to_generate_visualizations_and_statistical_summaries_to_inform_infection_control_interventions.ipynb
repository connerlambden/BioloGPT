Detailed analysis of NICU AMR prevalence data using pathogen count and resistance percentages.

```python
import pandas as pd
import plotly.express as px

data = {'Pathogen': ['CONS', 'Klebsiella pneumoniae', 'Acinetobacter spp.'],
        'Positive Cases': [70, 54, 23],
        'Positivity (%)': [16.7, 12.9, 5.6]}
df = pd.DataFrame(data)

fig = px.bar(df, x='Pathogen', y='Positivity (%)', color='Pathogen', title='AMR Pathogen Positivity (%) in NICU')
fig.show()
```

This notebook helps in visualizing key pathogen data and quantifying resistance levels to support targeted interventions.
