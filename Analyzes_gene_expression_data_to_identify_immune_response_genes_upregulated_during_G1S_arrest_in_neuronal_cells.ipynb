This notebook analyzes RNA-seq data to identify genes involved in immune responses that are upregulated during G1/S arrest.

```python
import pandas as pd
import numpy as np

# Load RNA-seq data
rna_seq_data = pd.read_csv('rna_seq_data.csv')

# Filter for immune response genes
immune_genes = rna_seq_data[rna_seq_data['gene_type'] == 'immune']

# Analyze expression levels during G1/S arrest
g1s_arrest_data = immune_genes[immune_genes['cell_cycle_phase'] == 'G1/S']

# Summary statistics
summary_stats = g1s_arrest_data.describe()
```
