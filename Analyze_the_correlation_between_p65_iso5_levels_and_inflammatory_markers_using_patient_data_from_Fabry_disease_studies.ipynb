This notebook will analyze the correlation between p65 iso5 levels and inflammatory markers in Fabry disease patients.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load patient data
patient_data = pd.read_csv('fabry_patient_data.csv')

# Analyze correlation
correlation = patient_data[['p65_iso5_levels', 'IL6_levels', 'TNF_alpha_levels']].corr()

# Visualize correlation
sns.heatmap(correlation, annot=True)
plt.title('Correlation between p65 iso5 levels and inflammatory markers')
plt.show()
```

This analysis provides insights into the relationship between p65 iso5 levels and inflammatory markers in Fabry disease patients.
