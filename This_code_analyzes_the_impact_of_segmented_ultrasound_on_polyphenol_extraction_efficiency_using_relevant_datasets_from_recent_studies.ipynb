### Analysis of Polyphenol Extraction Efficiency
This notebook will analyze the effects of segmented ultrasound on polyphenol extraction efficiency using data from various studies.

```python
import pandas as pd

# Load relevant datasets
datasets = {
    'apricot': {'polyphenol_yield': 10.8, 'method': 'Infrared'},
    'grape': {'polyphenol_yield': 4.8, 'method': 'Ultrasound'},
    'thai_plants': {'polyphenol_yield': 20.6, 'method': 'Microwave'}
}

# Create a DataFrame
extraction_data = pd.DataFrame(datasets).T
extraction_data
```

### Summary of Extraction Yields
The following table summarizes the polyphenol yields from different extraction methods.

```python
extraction_data.plot(kind='bar', y='polyphenol_yield', title='Polyphenol Yields by Extraction Method')
```
