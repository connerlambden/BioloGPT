### Vaccine Effectiveness Analysis
This notebook will guide you through analyzing vaccine effectiveness data using logistic regression.

```python
import pandas as pd
import statsmodels.api as sm

def calculate_ve(data):
    model = sm.Logit(data['case'], data[['vaccinated', 'age', 'health_status']])
    result = model.fit()
    odds_ratio = result.params['vaccinated']
    ve = 100 * (1 - odds_ratio)
    return ve

# Load your dataset
# data = pd.read_csv('vaccine_data.csv')
# ve = calculate_ve(data)
# print(f'Vaccine Effectiveness: {ve}%')
```
