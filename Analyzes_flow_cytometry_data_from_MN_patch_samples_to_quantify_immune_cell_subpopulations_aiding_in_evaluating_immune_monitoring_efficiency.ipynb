This notebook processes flow cytometry datasets from MN patch sampling experiments to identify and quantify T cell subsets and other immune cells.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load flow cytometry dataset
url = 'URL_to_flow_cytometry_data.csv'
data = pd.read_csv(url)

# Calculate counts of each immune cell type
cell_counts = data['cell_type'].value_counts()

# Plot immune cell type distribution
plt.figure(figsize=(10,6))
cell_counts.plot(kind='bar', color='#6A0C76')
plt.title('Immune Cell Type Distribution from MN Patch Samples')
plt.xlabel('Cell Type')
plt.ylabel('Count')
plt.show()
```

The code above loads the dataset, computes the frequency of each immune cell type, and visualizes the results. Further analysis compares these distributions with traditional sampling methods.

```python
import seaborn as sns

# Example comparison plot if dataset includes method labels
sns.barplot(x='method', y='T_cell_count', data=data, palette=['#6A0C76'])
plt.title('Comparison of T Cell Recovery Methods')
plt.xlabel('Sampling Method')
plt.ylabel('T Cell Count')
plt.show()
```

These plots help validate the effectiveness of MN patch sampling in recovering a specific immune cell profile compared to conventional techniques.
