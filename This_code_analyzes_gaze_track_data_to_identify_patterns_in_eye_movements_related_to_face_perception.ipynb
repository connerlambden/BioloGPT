### Gaze Track Data Analysis
This notebook analyzes gaze track data from the study to identify patterns in eye movements.

```python
import pandas as pd
import matplotlib.pyplot as plt

gaze_data = pd.read_csv('gaze_track_data.csv')
plt.figure(figsize=(10, 6))
plt.bar(gaze_data['Condition'], gaze_data['Accuracy'])
plt.title('Gaze Track Accuracy by Condition')
plt.xlabel('Condition')
plt.ylabel('Accuracy (%)')
plt.show()
```

### Discussion
This analysis provides insights into how gaze tracking accuracy varies across different conditions, contributing to our understanding of face perception.
