Analyse des données cliniques sur les kystes sébacés.

```python
import pandas as pd

# Charger les données cliniques sur les kystes sébacés
data = pd.read_csv('sebaceous_cysts_data.csv')

# Analyser les résultats des traitements
results = data.groupby('treatment').agg({'outcome': 'mean'})
results
```

Cette analyse permet d'identifier les traitements les plus efficaces pour les kystes sébacés.

```python
results.to_csv('treatment_outcomes.csv')
```
