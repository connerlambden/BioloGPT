### Mycotoxin Degradation Analysis
This notebook analyzes the degradation rates of mycotoxins using plasma treatment data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for mycotoxin degradation
data = {'Mycotoxin': ['AFB1', 'OTA', 'FB1'],
        'Initial Concentration (µg/kg)': [23.71, 164.66, 6850.25],
        'Final Concentration (15cm) (µg/kg)': [4.52, 26.13, 1123.44],
        'Degradation Percentage': [80.94, 84.07, 83.60]}

df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(df['Mycotoxin'], df['Degradation Percentage'], color='blue')
plt.title('Mycotoxin Degradation Percentage by Plasma Treatment')
plt.xlabel('Mycotoxin')
plt.ylabel('Degradation Percentage (%)')
plt.show()
```

This analysis provides insights into the effectiveness of plasma treatment in degrading mycotoxins.
