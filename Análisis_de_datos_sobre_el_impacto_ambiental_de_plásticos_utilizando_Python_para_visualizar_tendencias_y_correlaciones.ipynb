Análisis de datos sobre plásticos en la industria española.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Año': [2010, 2015, 2020], 'Plásticos (toneladas)': [100, 200, 300]}
df = pd.DataFrame(data)
plt.bar(df['Año'], df['Plásticos (toneladas)'])
plt.title('Aumento de Plásticos en la Industria Española')
plt.xlabel('Año')
plt.ylabel('Cantidad de Plásticos (toneladas)')
plt.show()
```

Este análisis muestra el aumento en el uso de plásticos en la industria española a lo largo de los años.
