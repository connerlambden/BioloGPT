### Morphometric Analysis of Triatominae Vectors
This notebook analyzes morphometric data to identify traits associated with insecticide resistance in Triatominae vectors.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load morphometric data
morphometric_data = pd.read_csv('morphometric_data.csv')

# Visualize the relationship between wing size and resistance
sns.scatterplot(data=morphometric_data, x='wing_size', y='resistance_level')
plt.title('Wing Size vs. Insecticide Resistance')
plt.xlabel('Wing Size (mm)')
plt.ylabel('Resistance Level')
plt.show()
```

### Discussion
This analysis helps identify key morphometric traits that correlate with insecticide resistance, providing insights for vector control strategies.

```python
# Save the plot
plt.savefig('wing_size_vs_resistance.png')
```
