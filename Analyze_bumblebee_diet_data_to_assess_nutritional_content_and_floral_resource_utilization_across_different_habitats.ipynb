This notebook will analyze pollen composition data from bumblebees collected in urban and rural settings.

```python
import pandas as pd

# Load dataset
dataset = pd.read_csv('bumblebee_diet_data.csv')

# Analyze nutritional content
nutritional_analysis = dataset.groupby('Location').agg({'Protein': 'mean', 'Carbohydrates': 'mean'})

# Display results
nutritional_analysis
```

The analysis provides insights into how land-use changes affect bumblebee nutrition.

```python
nutritional_analysis.to_csv('nutritional_analysis_results.csv')
```
