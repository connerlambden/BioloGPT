This notebook will analyze the correlation between training intensity and salivary oxytocin levels, as well as their impact on EGUS incidence.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load relevant datasets
oxytocin_data = pd.read_csv('oxytocin_levels.csv')
egus_data = pd.read_csv('egus_incidence.csv')

# Merge datasets on horse ID
merged_data = pd.merge(oxytocin_data, egus_data, on='horse_id')

# Analyze correlation
correlation = merged_data['oxytocin_level'].corr(merged_data['egus_incidence'])

# Plot results
plt.scatter(merged_data['oxytocin_level'], merged_data['egus_incidence'])
plt.title('Correlation between Oxytocin Levels and EGUS Incidence')
plt.xlabel('Oxytocin Level (pg/mL)')
plt.ylabel('EGUS Incidence')
plt.show()
```

This analysis will provide insights into how training intensity may influence oxytocin levels and the subsequent risk of EGUS.
