Analyze funding allocation data to assess EDI support.

```python
import pandas as pd

# Load funding data
data = pd.read_csv('funding_data.csv')

# Analyze disparities in funding allocation
summary = data.groupby('program').agg({'funding_amount': 'sum', 'diversity_score': 'mean'})
print(summary)
```
