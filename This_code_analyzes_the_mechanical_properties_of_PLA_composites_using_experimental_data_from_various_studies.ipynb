### Analysis of Mechanical Properties of PLA Composites
This notebook will analyze the mechanical properties of PLA composites based on data extracted from recent studies.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for PLA composites
data = {
    'Fiber Loading (%)': [10, 20, 30, 40],
    'Tensile Strength (MPa)': [41.54, 44.35, 47.28, 40.78],
    'Elongation at Break (%)': [1.96, 1.72, 1.46, 1.28]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(df['Fiber Loading (%)'], df['Tensile Strength (MPa)'], marker='o', label='Tensile Strength')
plt.plot(df['Fiber Loading (%)'], df['Elongation at Break (%)'], marker='o', label='Elongation at Break')
plt.title('Mechanical Properties of PLA Composites')
plt.xlabel('Fiber Loading (%)')
plt.ylabel('Mechanical Properties')
plt.legend()
plt.grid()
plt.show()
```
