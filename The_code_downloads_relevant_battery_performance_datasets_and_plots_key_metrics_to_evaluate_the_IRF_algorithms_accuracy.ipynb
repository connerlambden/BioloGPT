This notebook demonstrates data acquisition and visualization of battery health metrics.

```python
import pandas as pd
import plotly.express as px
# Load dataset from the provided URL
url = 'https://figshare.com/articles/dataset/eVTOL_Battery_Data/14226830/1'
df = pd.read_csv(url)
fig = px.bar(df, x='Metric', y='Value', title='Performance Metrics')
fig.show()
```
