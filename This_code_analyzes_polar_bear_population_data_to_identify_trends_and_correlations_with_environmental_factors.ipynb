### Polar Bear Population Analysis
This notebook analyzes population data of polar bears in relation to environmental changes.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load population data
data = {'Year': [1990, 2000, 2010, 2020, 2023], 'Kane_Basin': [224, 2650, 357, 3000, 357], 'Barents_Sea': [1900, 2200, 2400, 2600, 2650]}
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['Kane_Basin'], marker='o', label='Kane Basin Population')
plt.plot(df['Year'], df['Barents_Sea'], marker='o', label='Barents Sea Population')
plt.title('Polar Bear Population Trends')
plt.xlabel('Year')
plt.ylabel('Population Size')
plt.legend()
plt.grid()
plt.show()
```

### Discussion
This analysis provides insights into the population dynamics of polar bears and highlights the importance of conservation efforts.
