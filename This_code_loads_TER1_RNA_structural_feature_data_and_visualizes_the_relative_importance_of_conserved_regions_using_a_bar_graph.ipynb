This notebook demonstrates loading a dataset of TER1 RNA features, creating a DataFrame, and visualizing feature importance with Plotly.

```python
import pandas as pd
import plotly.express as px

# Load TER1 feature dataset
data = {'Feature': ['CS Regions', 'TWJ', 'Pseudoknot', 'Dispensable Regions'],
        'Importance': [8, 10, 9, 4]}
df = pd.DataFrame(data)

# Plot the relative importance of key structural elements
fig = px.bar(df, x='Feature', y='Importance', title='Relative Importance of TER1 RNA Structural Elements',
             labels={'Importance':'Importance Score', 'Feature':'Structural Element'},
             color='Feature', color_discrete_sequence=['#8A2BE2', '#6A0C76', '#6A0C76', '#C71585'])
fig.show()

print(df)
```

The bar chart clearly indicates that the TWJ is critically important, reinforcing the experimental findings of flexible scaffolding.
