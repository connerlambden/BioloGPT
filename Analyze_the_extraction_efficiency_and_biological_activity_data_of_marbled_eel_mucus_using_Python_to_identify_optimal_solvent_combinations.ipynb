This notebook will analyze the extraction efficiency and biological activity data of marbled eel mucus to identify optimal solvent combinations.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for extraction efficiency and biological activity
solvent_data = {
    'Solvent': ['Acetone', 'Ethanol', 'Methanol', 'Ethyl acetate', 'Chloroform', 'n-Hexane'],
    'RE%': [56.025, 23.375, 18.075, 18.45, 18.475, 17.8],
    'IC50_antioxidant': [36.603, 40.666, 46.984, 51.053, None, 56.898],
    'IC50_antibacterial_S.aureus': [65.0, None, None, None, None, None],
    'IC50_antibacterial_B.subtilis': [112.0, None, None, None, None, None]
}

# Create DataFrame
df = pd.DataFrame(solvent_data)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(df['Solvent'], df['RE%'], color='skyblue', label='Extraction Yield (%)')
plt.ylabel('Extraction Yield (%)')
plt.title('Extraction Yield of Marbled Eel Mucus by Solvent')
plt.legend()
plt.show()
```
