This notebook will analyze immune response data to Salmonella O antigen, focusing on IgA levels and T cell activation.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load immune response data
data = pd.read_csv('salmonella_immune_response.csv')

# Analyze IgA levels
plt.figure(figsize=(10, 6))
plt.bar(data['O_antigen_variant'], data['IgA_levels'])
plt.title('IgA Levels by O Antigen Variant')
plt.xlabel('O Antigen Variant')
plt.ylabel('IgA Levels')
plt.show()
```

The analysis reveals the relationship between different O antigen variants and the corresponding IgA levels.

```python
# Further analysis on T cell activation
plt.figure(figsize=(10, 6))
plt.scatter(data['O_antigen_variant'], data['T_cell_activation'], color='red')
plt.title('T Cell Activation by O Antigen Variant')
plt.xlabel('O Antigen Variant')
plt.ylabel('T Cell Activation')
plt.show()
```
