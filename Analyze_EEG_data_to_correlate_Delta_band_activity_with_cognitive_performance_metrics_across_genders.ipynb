This notebook will analyze EEG data focusing on Delta band activity and its correlation with cognitive performance metrics.

```python
import pandas as pd

data = pd.read_csv('eeg_data.csv')
# Analyze Delta band activity and cognitive performance metrics.
```

The analysis will include statistical tests to determine the significance of differences in Delta band activity between genders.

```python
import scipy.stats as stats

# Perform t-test between male and female Delta band activity
male_delta = data[data['gender'] == 'male']['delta_activity']
female_delta = data[data['gender'] == 'female']['delta_activity']
t_stat, p_value = stats.ttest_ind(male_delta, female_delta)
```
