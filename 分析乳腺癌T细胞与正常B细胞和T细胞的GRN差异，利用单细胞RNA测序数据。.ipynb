### 数据准备
从公共数据库下载乳腺癌T细胞和正常细胞的单细胞RNA测序数据。

```python
import scanpy as sc

# 加载数据
adata_cancer = sc.read('breast_cancer_t_cells.h5ad')
data_normal = sc.read('normal_b_t_cells.h5ad')

# 数据预处理
sc.pp.normalize_total(adata_cancer, target_sum=1e4)
sc.pp.log1p(adata_cancer)
sc.pp.normalize_total(data_normal, target_sum=1e4)
sc.pp.log1p(data_normal)
```
