This notebook will analyze RNA-Seq data from hTERT-HPNE cells treated with various stressors to identify differentially expressed genes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('hTERT_HPNE_stress_data.csv')

# Analyze differential expression
# (Assuming the dataset has columns 'gene', 'expression', 'condition')
diff_expr = data[data['condition'] == 'stressed']

# Visualize the results
plt.figure(figsize=(10,6))
sns.barplot(x='gene', y='expression', data=diff_expr)
plt.title('Differential Gene Expression in hTERT-HPNE Cells Under Stress')
plt.xticks(rotation=90)
plt.show()
```

This analysis will help identify which genes are significantly upregulated or downregulated in response to stress, providing insights into the mechanisms of pancreatitis.

```python
# Further analysis can include statistical tests to validate the findings
from scipy import stats

# Example: t-test for a specific gene
gene_data = data[data['gene'] == 'specific_gene']
t_stat, p_value = stats.ttest_ind(gene_data[gene_data['condition'] == 'stressed']['expression'],
gene_data[gene_data['condition'] == 'control']['expression'])

print(f'T-statistic: {t_stat}, P-value: {p_value}')
```
