### Binding Affinity Analysis of Vedolizumab

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for binding affinities
data = {'Target': ['α4β7 Integrin', 'MAdCAM-1', 'Memory T Cells'], 'IC50 (nM)': [0.3, 0.225, 0.3]}
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Target'], df['IC50 (nM)'], color='blue')
plt.title('Vedolizumab Binding Affinity')
plt.xlabel('Targets')
plt.ylabel('IC50 (nM)')
plt.show()
```

This analysis visualizes the binding affinities of vedolizumab to its targets, aiding in understanding its selective action.
