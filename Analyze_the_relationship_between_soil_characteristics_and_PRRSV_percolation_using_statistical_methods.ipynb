This notebook will analyze the influence of soil characteristics on PRRSV percolation rates.

```python
import pandas as pd
import statsmodels.api as sm

# Load soil characteristics data
data = pd.read_csv('soil_characteristics.csv')

# Perform regression analysis to assess the impact of soil pH and organic matter on PRRSV percolation rates
X = data[['pH', 'Organic_Matter']]
Y = data['PRRSV_Percolation_Rate']
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
model.summary()
```

The regression analysis results will provide insights into how soil characteristics influence PRRSV percolation.

```python
# Display the results
print(model.summary())
```
