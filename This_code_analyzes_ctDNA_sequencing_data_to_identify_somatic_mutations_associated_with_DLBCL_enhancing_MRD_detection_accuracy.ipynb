### ctDNA Analysis for DLBCL MRD Detection
This notebook will guide you through the analysis of ctDNA sequencing data to identify somatic mutations relevant to DLBCL.

```python
import pandas as pd

# Load ctDNA sequencing data
data = pd.read_csv('ctDNA_data.csv')

# Analyze mutation frequency
mutation_counts = data['mutation'].value_counts()
mutation_counts
```

### Results
The mutation frequency analysis will help identify prevalent mutations in DLBCL patients, aiding in MRD detection.

```python
# Visualize mutation frequency
import matplotlib.pyplot as plt
mutation_counts.plot(kind='bar')
plt.title('Mutation Frequency in DLBCL ctDNA')
plt.xlabel('Mutations')
plt.ylabel('Frequency')
plt.show()
```
