전압과 전류를 입력받아 전력을 계산하는 간단한 코드입니다.

```python
# 전압과 전류를 입력받아 전력을 계산하는 함수
def calculate_power(voltage, current):
    return voltage * current

# 예시 사용
voltage = 10  # Volts
current = 2    # Amperes
power = calculate_power(voltage, current)
print(f'전력: {power} W')
```
