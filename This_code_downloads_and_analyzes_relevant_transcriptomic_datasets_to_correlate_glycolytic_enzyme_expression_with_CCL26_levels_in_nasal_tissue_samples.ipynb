Load and preprocess the transcriptomic dataset from ECRSwNP patients to examine correlations between glycolysis-related genes and CCL26 expression.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load dataset (replace 'data.csv' with the actual dataset path)
data = pd.read_csv('data.csv')
# Correlate glycolytic enzyme expression (HK2, PFKL, PKM2, LDHA) with CCL26
genes = ['HK2', 'PFKL', 'PKM2', 'LDHA', 'CCL26']
data_subset = data[genes]
correlation_matrix = data_subset.corr()
# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis')
plt.title('Correlation between Glycolytic Genes and CCL26')
plt.show()
```

The above code visualizes the relationships between key glycolytic enzymes and CCL26 expression, aiding hypothesis testing on metabolic regulation in ECRSwNP.
