### Data Analysis of Thickened Infant Formula Viscosity
This analysis will explore the relationship between thickener addition and viscosity in infant formulas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for thickener addition and viscosity
thickener_data = {
    'Thickener Addition (g/100g)': [2, 3, 4],
    'Mean Viscosity (mPa.s)': [11.3, 28.44, 65.01]
}

# Create DataFrame
df = pd.DataFrame(thickener_data)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(df['Thickener Addition (g/100g)'], df['Mean Viscosity (mPa.s)'], marker='o')
plt.title('Viscosity of Thickened Infant Formula')
plt.xlabel('Thickener Addition (g/100g)')
plt.ylabel('Mean Viscosity (mPa.s)')
plt.grid()
plt.show()
```
