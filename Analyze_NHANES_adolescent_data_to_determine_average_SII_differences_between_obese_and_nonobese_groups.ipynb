Load and filter the NHANES dataset for adolescents aged 12-19 years, then compute the average SII for obese versus non-obese groups.

```python
import pandas as pd
# Load dataset
nhanes = pd.read_csv('nhanes_adolescents.csv')
adolescents = nhanes[(nhanes['age'] >= 12) & (nhanes['age'] <= 19)]
# Group by obesity status and compute mean SII
group_stats = adolescents.groupby('obesity_status')['SII'].mean()
print(group_stats)
```

Generate a bar plot using Plotly to visually compare average SII between group categories.

```python
import plotly.express as px
fig = px.bar(x=group_stats.index, y=group_stats.values, labels={'x':'Obesity Status','y':'Average SII'}, title='Average SII in Obese vs Non-Obese Adolescents')
fig.show()
```
