This notebook retrieves relevant yeast gene expression data and integrates survival assay metrics to correlate gene knockouts with freeze–thaw tolerance.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://example.com/yeast_freezethaw_data.csv')
df['viability_change'] = df['post_freeze_viability'] - df['control_viability']
summary = df.groupby('gene')['viability_change'].mean()
print(summary)
```

The above code succinctly demonstrates data downloading, processing, and summary statistics for each gene knockout relative to freeze–thaw survival.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.barplot(x=summary.index, y=summary.values, palette='viridis')
plt.title('Average Viability Change by Gene Knockout')
plt.xlabel('Gene')
plt.ylabel('Average Viability Change')
plt.show()
```
