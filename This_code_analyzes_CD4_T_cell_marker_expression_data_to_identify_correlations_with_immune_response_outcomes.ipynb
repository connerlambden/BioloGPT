### CD4 T Cell Marker Analysis
This notebook analyzes expression data of CD4 T cell markers to identify patterns associated with immune responses.

```python
import pandas as pd

# Load CD4 T cell marker expression data
data = pd.read_csv('cd4_markers_expression.csv')

# Analyze correlations between markers and immune response outcomes
correlation_matrix = data.corr()
```

### Correlation Analysis
The correlation matrix provides insights into how different CD4 T cell markers relate to each other and to immune response metrics.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('CD4 T Cell Marker Correlation Matrix')
plt.show()
```
