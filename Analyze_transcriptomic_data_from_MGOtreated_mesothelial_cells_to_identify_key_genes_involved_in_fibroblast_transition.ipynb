This notebook will analyze transcriptomic data to identify genes upregulated in MGO-treated mesothelial cells.

```python
import pandas as pd
# Load transcriptomic data
transcriptomic_data = pd.read_csv('mgo_transcriptomic_data.csv')
# Analyze upregulated genes
upregulated_genes = transcriptomic_data[transcriptomic_data['log2FoldChange'] > 1]
```
