This notebook simulates the dynamics of HGT in bacterial populations with and without engineered plasmids.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate HGT dynamics
# Parameters: population size, initial resistance, transfer rate
population_size = 1000
initial_resistance = 0.01
transfer_rate = 0.1

# Simulate generations
for generation in range(100):
    # Calculate new resistance
    new_resistance = initial_resistance * (1 + transfer_rate)
    initial_resistance = new_resistance

plt.plot(range(100), initial_resistance)
plt.title('HGT Dynamics Over Generations')
plt.xlabel('Generations')
plt.ylabel('Resistance Rate')
plt.show()
```

This simulation provides insights into how resistance rates change over generations with varying transfer rates.
