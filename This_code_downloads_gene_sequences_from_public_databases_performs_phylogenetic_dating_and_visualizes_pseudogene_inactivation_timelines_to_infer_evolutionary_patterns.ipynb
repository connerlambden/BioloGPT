Download and preprocess genomic data from established datasets on myrmecophagous mammals. Identify pseudogenes and extract coding regions.

```python
import requests
import pandas as pd
# Download data from Zenodo
url = 'https://zenodo.org/record/14895250'
data = requests.get(url).json()
df = pd.DataFrame(data)
# Process and filter dataset for pseudogene analysis
print(df.head())
```

Perform phylogenetic analysis using available tools like PAML and visualize the timing of gene inactivation events.

```python
import matplotlib.pyplot as plt
# Example data for gene inactivation times (Mya)
genes = ['MYH16', 'DSPP', 'TAS1R1', 'TAS1R3', 'PKD2L1']
inactivation_times = [60, 50, 45, 30, 10]
plt.figure(figsize=(10,5))
plt.scatter(inactivation_times, genes, color='#6A0C76')
plt.xlabel('Millions of Years Ago')
plt.ylabel('Gene')
plt.title('Pseudogene Inactivation Timeline')
plt.gca().invert_xaxis()
plt.show()
```
