This notebook will analyze environmental cleanliness data collected from NICUs, focusing on its correlation with infection rates.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
cleanliness_data = pd.read_csv('cleanliness_data.csv')

# Analyze the correlation between cleanliness scores and infection rates
correlation = cleanliness_data.corr()
sns.heatmap(correlation, annot=True)
plt.title('Correlation between Cleanliness and Infection Rates')
plt.show()
```

The analysis will provide insights into how environmental cleanliness impacts infection rates, guiding future interventions.
