This notebook will analyze diaper rash prevalence data, focusing on risk factors such as moisture exposure and treatment outcomes.

```python
import pandas as pd

# Load diaper rash data
data = pd.read_csv('diaper_rash_data.csv')

# Analyze prevalence by age group
age_group_analysis = data.groupby('age_group')['prevalence'].mean()

# Display results
age_group_analysis
```

The analysis reveals the average prevalence of diaper rash across different age groups, helping to identify vulnerable populations.

```python
# Visualize the results
import matplotlib.pyplot as plt

plt.bar(age_group_analysis.index, age_group_analysis.values)
plt.title('Diaper Rash Prevalence by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Prevalence (%)')
plt.show()
```
