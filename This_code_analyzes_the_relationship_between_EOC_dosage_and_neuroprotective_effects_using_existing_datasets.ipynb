### Data Analysis of EOC and OTA Interaction

```python
import pandas as pd

data = {'EOC_Dose': [0, 0.01, 0.1, 1, 10, 100], 'Cell_Viability': [40, 68, 76, 83, 89, 95]}
df = pd.DataFrame(data)

# Plotting the data
import matplotlib.pyplot as plt
plt.plot(df['EOC_Dose'], df['Cell_Viability'], marker='o')
plt.title('Effect of EOC Dose on Cell Viability')
plt.xlabel('EOC Dose (mg/kg)')
plt.ylabel('Cell Viability (%)')
plt.xscale('log')
plt.grid()
plt.show()
```

This analysis visualizes the relationship between EOC dosage and cell viability, indicating a potential dose-dependent effect.
