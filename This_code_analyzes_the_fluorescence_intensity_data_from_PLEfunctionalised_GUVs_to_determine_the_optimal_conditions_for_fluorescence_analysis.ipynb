### Data Analysis of Fluorescence Intensity in PLE-Functionalised GUVs
This notebook will analyze fluorescence intensity data to optimize conditions for using PLE-functionalised GUVs.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load fluorescence intensity data
data = pd.read_csv('fluorescence_data.csv')

# Plotting the fluorescence intensity
plt.figure(figsize=(10, 6))
plt.plot(data['Condition'], data['Intensity'], marker='o')
plt.title('Fluorescence Intensity of PLE-Functionalised GUVs')
plt.xlabel('Condition')
plt.ylabel('Fluorescence Intensity')
plt.grid()
plt.show()
```
