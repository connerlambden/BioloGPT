### Analysis of ATR Mutations on DNA Replication Timing
This notebook will analyze the MFA-seq data to assess how ATR mutations affect DNA replication timing in Leishmania major.

```python
import pandas as pd

# Load MFA-seq data
mfa_data = pd.read_csv('mfa_seq_data.csv')

# Analyze replication timing based on ATR mutation status
replication_timing = mfa_data.groupby('ATR_mutation_status')['replication_timing'].mean()
print(replication_timing)
```
