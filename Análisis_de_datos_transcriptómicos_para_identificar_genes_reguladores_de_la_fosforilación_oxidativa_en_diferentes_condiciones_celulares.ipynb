Este análisis busca correlacionar la expresión génica con la actividad de la fosforilación oxidativa en células sanas y cancerosas.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar datos de expresión génica
data = pd.read_csv('gene_expression_data.csv')

# Análisis de correlación
correlation = data.corr()

# Visualización de la matriz de correlación
sns.heatmap(correlation, annot=True)
plt.title('Matriz de Correlación de Expresión Génica')
plt.show()
```

Este análisis permite identificar patrones de expresión que pueden estar relacionados con la regulación de la fosforilación oxidativa.

```python
# Filtrar genes relacionados con la fosforilación oxidativa
genes_oxphos = data[data['pathway'] == 'oxidative phosphorylation']

# Guardar resultados
genes_oxphos.to_csv('oxphos_genes.csv', index=False)
```
