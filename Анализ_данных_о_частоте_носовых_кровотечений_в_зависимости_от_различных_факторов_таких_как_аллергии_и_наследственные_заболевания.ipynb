Анализ данных о частоте носовых кровотечений в зависимости от различных факторов.

```python
import pandas as pd

# Загрузка данных о носовых кровотечениях
data = pd.read_csv('nosebleeds_data.csv')

# Анализ частоты носовых кровотечений по факторам
frequency_analysis = data.groupby('factor').mean()
```

Анализ частоты носовых кровотечений по различным факторам.

```python
frequency_analysis.to_csv('frequency_analysis.csv')
```
