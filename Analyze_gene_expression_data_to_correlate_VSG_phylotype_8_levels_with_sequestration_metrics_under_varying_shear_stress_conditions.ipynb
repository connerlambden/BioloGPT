This notebook will analyze RNA-seq data to identify correlations between VSG phylotype 8 expression and T. congolense sequestration.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load gene expression data
expression_data = pd.read_csv('vsg_expression_data.csv')

# Analyze correlation between VSG phylotype 8 and sequestration metrics
correlation = expression_data.corr()

# Visualize the correlation matrix
sns.heatmap(correlation, annot=True)
plt.title('Correlation between VSG Expression and Sequestration')
plt.show()
```

This analysis will help identify key relationships between VSG expression and the ability of T. congolense to sequester.
