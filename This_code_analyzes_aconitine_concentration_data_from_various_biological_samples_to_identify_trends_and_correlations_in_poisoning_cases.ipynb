### Aconitine Concentration Analysis
This notebook analyzes aconitine concentration data from post-mortem specimens to identify trends.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for aconitine concentrations
data = {
    'Specimen': ['Peripheral Blood', 'Heart Blood', 'Urine', 'Stomach Content', 'Bile', 'Vitreous Humor', 'Pericardial Fluid', 'Liver'],
    'Concentration (µg/L)': [14.1, 51.4, 399, 2500, 189, 8.4, 54.7, 492]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10,6))
plt.bar(df['Specimen'], df['Concentration (µg/L)'], color='purple')
plt.title('Aconitine Concentration in Various Specimens')
plt.xlabel('Specimen Type')
plt.ylabel('Concentration (µg/L)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Discussion
This analysis provides insights into the concentration of aconitine across different biological specimens, highlighting the most affected matrices.
