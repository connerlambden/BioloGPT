This notebook simulates genetic drift in a population over multiple generations.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate genetic drift
population_size = 100
allele_frequency = 0.5
num_generations = 50
frequencies = []

for _ in range(num_generations):
    allele_frequency += np.random.normal(0, 0.1)  # Random drift
    allele_frequency = max(0, min(1, allele_frequency))  # Keep within bounds
    frequencies.append(allele_frequency)

plt.plot(frequencies)
plt.title('Genetic Drift Simulation')
plt.xlabel('Generations')
plt.ylabel('Allele Frequency')
plt.show()
```

This simulation demonstrates how allele frequencies can fluctuate due to genetic drift.

```python
# Final plot of allele frequencies over generations
plt.plot(frequencies)
plt.title('Genetic Drift Over Generations')
plt.xlabel('Generations')
plt.ylabel('Allele Frequency')
plt.show()
```

This analysis provides insights into the stochastic nature of genetic drift.
