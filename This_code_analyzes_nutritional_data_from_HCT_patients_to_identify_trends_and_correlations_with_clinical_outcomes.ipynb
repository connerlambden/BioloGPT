### Nutritional Data Analysis for HCT Patients
This notebook analyzes the nutritional status of patients undergoing HCT, focusing on BMI, MUST, and GLIM scores.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Time Point': ['T0', 'T1', 'T2', 'T3'], 'Mean BMI': [25.3, 24.0, 23.8, 23.8], 'MUST Score': [0.3, 1.16, 1.89, 1.59]}
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(df['Time Point'], df['Mean BMI'], marker='o', label='Mean BMI')
plt.plot(df['Time Point'], df['MUST Score'], marker='o', label='MUST Score')
plt.title('Nutritional Status Over Time')
plt.xlabel('Time Points')
plt.ylabel('Scores')
plt.legend()
plt.show()
```

### Discussion
This analysis provides insights into the nutritional trends observed in HCT patients, highlighting the need for targeted interventions.
