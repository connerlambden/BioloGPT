This notebook analyzes the pharmacokinetics of morphine based on various administration routes.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for Cmax values
routes = ['Intraveineuse', 'Orale', 'Sublinguale', 'Buccale', 'Buccale lente']
Cmax_values = [273.2, 20.9, 19.3, 15.5, 11.5]

# Create DataFrame
data = pd.DataFrame({'Route': routes, 'Cmax (nmol/L)': Cmax_values})

# Plotting
plt.bar(data['Route'], data['Cmax (nmol/L)'])
plt.title('Cmax de la morphine selon la voie d\u2019administration')
plt.xlabel('Voie d\u2019administration')
plt.ylabel('Cmax (nmol/L)')
plt.show()
```

This analysis provides insights into the pharmacokinetics of morphine, aiding in optimizing pain management strategies.
