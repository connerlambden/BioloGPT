This section loads the ME/CFS and Long COVID datasets, performs a combined analysis of their metabolic flux profiles, and visualizes key differences using a bar graph.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load datasets from provided sources (replace with actual file paths)
me_cfs_data = pd.read_csv('GSE245661.csv')  # ME/CFS muscle tissue
long_covid_data = pd.read_csv('MTBLS9103.csv')  # Long COVID metabolomics

# Concatenate datasets for a unified view
merged_data = pd.concat([me_cfs_data, long_covid_data])

diff_flux = merged_data.groupby('reaction')['flux'].mean().reset_index()
# Identify top 10 reactions with highest mean flux
top10 = diff_flux.sort_values(by='flux', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top10['reaction'], top10['flux'], color='#6A0C76')
plt.xticks(rotation=45)
plt.title('Top 10 Metabolic Reactions by Average Flux')
plt.xlabel('Reaction')
plt.ylabel('Average Flux')
plt.tight_layout()
plt.show()
```

The code above visualizes the key reactions with altered flux, providing insights into the metabolic imbalances in these conditions.

```python
from scipy import stats

# Example statistical test for a specific reaction
reaction_of_interest = 'alanine_aspartate_metabolism'
flux_me_cfs = me_cfs_data[me_cfs_data['reaction'] == reaction_of_interest]['flux']
flux_long_covid = long_covid_data[long_covid_data['reaction'] == reaction_of_interest]['flux']
stat, p_val = stats.ttest_ind(flux_me_cfs, flux_long_covid)
print(f'T-test for {reaction_of_interest}: statistic = {stat}, p-value = {p_val}')
```

These analyses help test the significance of dysregulation in target metabolic pathways, supporting therapeutic hypotheses.
