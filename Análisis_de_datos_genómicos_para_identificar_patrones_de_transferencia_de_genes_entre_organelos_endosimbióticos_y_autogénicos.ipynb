Este análisis busca correlacionar la presencia de genes específicos en organelos endosimbióticos y autogénicos.

```python
import pandas as pd

# Cargar datos de genes de organelos
genes = pd.read_csv('genes_organelos.csv')

# Análisis de patrones de transferencia
genes['transfer_pattern'] = genes.apply(lambda x: 'endosimbiótico' if x['type'] == 'mitocondria' else 'autogénico', axis=1)

# Resultados
print(genes.head())
```

Este código permite identificar y clasificar genes según su origen en organelos endosimbióticos o autogénicos.

```python
# Visualización de resultados
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
genes['transfer_pattern'].value_counts().plot(kind='bar')
plt.title('Patrones de Transferencia de Genes')
plt.xlabel('Tipo de Organelo')
plt.ylabel('Número de Genes')
plt.show()
```
