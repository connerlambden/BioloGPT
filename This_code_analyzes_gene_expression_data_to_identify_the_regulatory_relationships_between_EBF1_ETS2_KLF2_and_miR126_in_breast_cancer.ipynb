### Gene Expression Analysis of EBF1, ETS2, KLF2, and miR-126 in Breast Cancer

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load gene expression data
expression_data = pd.read_csv('breast_cancer_expression.csv')

# Filter for relevant genes
relevant_genes = ['EBF1', 'ETS2', 'KLF2', 'miR-126']
filtered_data = expression_data[expression_data['gene'].isin(relevant_genes)]

# Plot the expression levels
plt.figure(figsize=(10, 6))
sns.boxplot(x='gene', y='expression', data=filtered_data)
plt.title('Expression Levels of EBF1, ETS2, KLF2, and miR-126 in Breast Cancer')
plt.show()
```
