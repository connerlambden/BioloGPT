### Analysis of Ancient Genomic Data
This notebook will explore the genomic sequences of Yersinia pestis from ancient sheep and humans.

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load genomic data
data = pd.read_csv('ancient_genomic_data.csv')

# Analyze the presence of Yersinia pestis in different hosts
results = data[data['pathogen'] == 'Yersinia pestis']
results.head()
```

### Discussion
This analysis will help identify the transmission dynamics of Yersinia pestis across different hosts.

```python
# Further analysis and visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(results['host'], bins=20)
plt.title('Distribution of Yersinia pestis in Hosts')
plt.xlabel('Host')
plt.ylabel('Frequency')
plt.show()
```
