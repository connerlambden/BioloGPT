### Data Preparation
Load genetic and lifestyle datasets from dementia cohorts to prepare for analysis.

```python
import pandas as pd
# Load datasets
genetic_data = pd.read_csv('genetic_data.csv')
lifestyle_data = pd.read_csv('lifestyle_data.csv')
```

### Data Analysis
Merge datasets and perform statistical analysis to identify significant predictors of cognitive resilience.

```python
merged_data = pd.merge(genetic_data, lifestyle_data, on='participant_id')
# Perform analysis
results = merged_data.groupby('genetic_risk').agg({'cognitive_score':'mean'})
```

### Visualization
Visualize the results to interpret the impact of genetic and lifestyle factors on cognitive resilience.

```python
import matplotlib.pyplot as plt
plt.bar(results.index, results['cognitive_score'])
plt.title('Cognitive Resilience by Genetic Risk')
plt.xlabel('Genetic Risk')
plt.ylabel('Average Cognitive Score')
plt.show()
```
