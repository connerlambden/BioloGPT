### Data Analysis of Retinal and Plasma Biomarkers
This notebook analyzes the correlation between retinal and plasma biomarkers for Alzheimer's disease detection.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data for analysis
biomarkers = {'Biomarker': ['p-tau217', 'Putative retinal gliosis', 'Inner inferior RNFL thickness', 'Inner superior RNFL thickness', 'Inner nasal RNFL thickness'],
              'Mean Value': [0.69, 1.0, 24, 24, 20]}
df = pd.DataFrame(biomarkers)

# Correlation analysis
correlation = df.corr()

# Visualization
sns.heatmap(correlation, annot=True)
plt.title('Correlation between Biomarkers')
plt.show()
```
