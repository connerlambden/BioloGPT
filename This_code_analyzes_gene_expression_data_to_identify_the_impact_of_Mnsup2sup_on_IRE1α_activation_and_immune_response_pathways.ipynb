Analyze RNA-Seq data from cancer models treated with Mn<sup>2+</sup> to assess changes in IRE1α and immune-related gene expression.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_data.csv')
sns.boxplot(x='treatment', y='IRE1A_expression', data=data)
plt.title('IRE1A Expression by Treatment')
plt.show()
```

This analysis visualizes the expression levels of IRE1α in response to Mn<sup>2+</sup> treatment, providing insights into its role in immunotherapy.
