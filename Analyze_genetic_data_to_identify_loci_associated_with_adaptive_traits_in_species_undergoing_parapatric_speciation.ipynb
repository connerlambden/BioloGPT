This notebook will analyze genetic data from species undergoing parapatric speciation to identify adaptive loci.

```python
import pandas as pd
import numpy as np

# Load genetic data
genetic_data = pd.read_csv('genetic_data.csv')

# Analyze loci associated with adaptive traits
adaptive_loci = genetic_data[genetic_data['trait'] == 'adaptive']

# Summary statistics
summary = adaptive_loci.describe()
```

The analysis provides insights into the genetic basis of adaptation in parapatric populations.

```python
summary.to_csv('adaptive_loci_summary.csv')
```
