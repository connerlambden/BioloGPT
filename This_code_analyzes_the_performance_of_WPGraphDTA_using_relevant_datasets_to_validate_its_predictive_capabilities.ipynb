### Data Preparation
We will load the datasets used in the WPGraphDTA study to analyze its performance.

```python
import pandas as pd

# Load datasets
kiba_data = pd.read_csv('kiba_dataset.csv')
davis_data = pd.read_csv('davis_dataset.csv')

# Display the first few rows of the datasets
print(kiba_data.head())
print(davis_data.head())
```

### Model Evaluation
We will evaluate the performance of WPGraphDTA on the Kiba and Davis datasets.

```python
from sklearn.metrics import mean_squared_error, r2_score

# Assuming predictions and actual values are available
predictions_kiba = kiba_data['predicted_values']
actual_kiba = kiba_data['actual_values']

mse_kiba = mean_squared_error(actual_kiba, predictions_kiba)
r2_kiba = r2_score(actual_kiba, predictions_kiba)

print(f'MSE for Kiba: {mse_kiba}, R2 Score: {r2_kiba}')
```
