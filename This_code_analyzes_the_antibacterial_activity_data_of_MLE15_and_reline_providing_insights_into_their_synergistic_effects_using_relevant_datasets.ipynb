### Data Analysis of Antibacterial Activity
This notebook analyzes the antibacterial activity of MLE-15 and reline against various bacterial strains.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load antibacterial activity data
activity_data = pd.read_csv('antibacterial_activity.csv')

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(activity_data['Treatment'], activity_data['MIC'])
plt.title('Antibacterial Activity of MLE-15 and Reline')
plt.xlabel('Treatment')
plt.ylabel('MIC (µg/mL)')
plt.show()
```

### Discussion
This analysis provides a visual representation of the antibacterial activity of MLE-15 and reline, highlighting their potential as a combined treatment.
