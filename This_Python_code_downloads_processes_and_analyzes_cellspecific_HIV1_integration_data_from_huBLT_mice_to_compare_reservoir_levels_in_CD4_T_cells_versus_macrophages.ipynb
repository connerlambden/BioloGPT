This notebook outlines the analysis of HIV-DNA integration levels in sorted cell populations from the hu-BLT mouse model using publicly available dataset URL_0.

```python
import pandas as pd
# Code to download and parse the HIV-DNA integration dataset for CD4+ T cells and macrophages
integration_data = pd.read_csv('URL_0/integration_data.csv')
summary = integration_data.groupby('Cell_Type')['HIV_DNA_copies'].mean()
print(summary)
```

The above code computes the average proviral load in CD4+ T cells and macrophages, highlighting the differences discussed in the study.

```python
import plotly.express as px
fig = px.bar(summary.reset_index(), x='Cell_Type', y='HIV_DNA_copies', title='Average HIV-DNA Copies in Sorted Cells')
fig.show()
```

This visualization confirms that CD4+ T cells harbor significantly more integrated HIV-DNA than macrophages during ART.
