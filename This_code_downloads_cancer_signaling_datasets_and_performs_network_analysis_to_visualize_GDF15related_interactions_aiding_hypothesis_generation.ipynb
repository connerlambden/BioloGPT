Load and preprocess curated datasets on GDF15 interactions from the provided Excel files.

```python
import pandas as pd

data = pd.read_excel('Online_Resource_1.xlsx')
print(data.head())
```

Build a network graph of GDF15-related molecules and visualize using Plotly.

```python
import plotly.graph_objects as go

nodes = data['Molecule'].unique()

fig = go.Figure(data=[go.Scatter(x=[i for i in range(len(nodes))], y=[1]*len(nodes), mode='markers+text', text=nodes)])
fig.update_layout(title='GDF15 Interaction Network', showlegend=False)
fig.show()
```
