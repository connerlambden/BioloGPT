### BDNF Levels Analysis in Psoriasis Patients
This notebook analyzes BDNF levels from ELISA data to explore correlations with treatment outcomes.

```python
import pandas as pd

# Load BDNF data
bdnf_data = pd.DataFrame({
    'Group': ['Healthy Controls', 'Patients (Before Therapy)', 'Patients (After Therapy)'],
    'BDNF Levels (ng/mL)': [25.14, 72.49, 63.60]
})

# Display the data
bdnf_data
```

### Statistical Analysis
Perform statistical tests to compare BDNF levels across groups.

```python
from scipy import stats

# Perform ANOVA
f_statistic, p_value = stats.f_oneway(bdnf_data['BDNF Levels (ng/mL)'][bdnf_data['Group'] == 'Healthy Controls'],
                                        bdnf_data['BDNF Levels (ng/mL)'][bdnf_data['Group'] == 'Patients (Before Therapy)'],
                                        bdnf_data['BDNF Levels (ng/mL)'][bdnf_data['Group'] == 'Patients (After Therapy)'])

f_statistic, p_value
```
