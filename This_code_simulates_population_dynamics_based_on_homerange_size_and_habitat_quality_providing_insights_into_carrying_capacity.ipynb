### Population Dynamics Simulation
This notebook simulates the effects of home-range size on population dynamics in fragmented landscapes.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
home_range_sizes = [1, 2, 3]  # Example home range sizes
habitat_quality = np.linspace(0, 1, 100)  # Habitat quality from 0 to 1

# Simulate carrying capacity
carrying_capacity = {size: 100 * (1 - (habitat_quality / size)) for size in home_range_sizes}

# Plotting
plt.figure(figsize=(10, 6))
for size, capacity in carrying_capacity.items():
    plt.plot(habitat_quality, capacity, label=f'Home Range Size: {size}')
plt.title('Carrying Capacity vs Habitat Quality')
plt.xlabel('Habitat Quality')
plt.ylabel('Carrying Capacity')
plt.legend()
plt.grid()
plt.show()
```

### Discussion
This simulation illustrates how carrying capacity varies with habitat quality and home-range size, supporting the findings of the paper.
