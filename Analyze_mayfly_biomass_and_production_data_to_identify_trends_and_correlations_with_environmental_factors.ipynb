### Mayfly Biomass and Production Analysis
This notebook analyzes the biomass and production data of mayflies from the study, focusing on the differences between forested and deforested areas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Species': ['Xyrodromeus', 'D. merina', 'X. namarona'],
        'Biomass_Savoka': [4.5, 4.5, 1],
        'Biomass_Forest': [1, 1, 1]}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Species'], df['Biomass_Savoka'], color='orange', label='Savoka')
plt.bar(df['Species'], df['Biomass_Forest'], color='green', label='Forest', alpha=0.5)
plt.title('Mayfly Biomass Comparison')
plt.ylabel('Biomass (times greater)')
plt.legend()
plt.show()
```
