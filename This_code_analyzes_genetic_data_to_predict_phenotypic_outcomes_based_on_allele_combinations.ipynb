### Genetic Cross Analysis
This notebook will analyze the genetic cross between two plant genotypes to predict the phenotypic ratios of the offspring.

```python
# Define parental genotypes
parent1 = {'T': 1, 't': 1, 'P': 1, 'p': 1}  # TtPp
parent2 = {'t': 2, 'p': 2}  # ttpp

# Calculate offspring phenotypes
# Initialize phenotype counts
phenotypes = {'Tall Purple': 0, 'Tall White': 0, 'Short Purple': 0, 'Short White': 0}

# Simulate Punnett square outcomes
for gamete1 in parent1.keys():
    for gamete2 in parent2.keys():
        if gamete1 == 'T' and gamete2 == 't':
            phenotypes['Tall Purple'] += 1
        elif gamete1 == 'T' and gamete2 == 'p':
            phenotypes['Tall White'] += 1
        elif gamete1 == 't' and gamete2 == 'p':
            phenotypes['Short Purple'] += 1
        elif gamete1 == 't' and gamete2 == 'p':
            phenotypes['Short White'] += 1

# Calculate percentages
total_offspring = sum(phenotypes.values())
short_white_percentage = (phenotypes['Short White'] / total_offspring) * 100
short_white_percentage
```
