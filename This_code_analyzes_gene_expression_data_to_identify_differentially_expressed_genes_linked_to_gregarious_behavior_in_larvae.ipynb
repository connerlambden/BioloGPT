### Gene Expression Analysis for Gregarious Behavior

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load gene expression data
expression_data = pd.read_csv('gene_expression_data.csv')

# Identify differentially expressed genes (DEGs)
deg_threshold = 0.05
results = stats.ttest_ind(expression_data['gregarious'], expression_data['solitary'])

# Filter DEGs
degs = expression_data[results.pvalue < deg_threshold]
```

### Summary of Results
This analysis identifies genes that are significantly differentially expressed between gregarious and solitary larvae, providing insights into the molecular basis of social behavior.

```python
# Display DEGs
print(degs)
```
