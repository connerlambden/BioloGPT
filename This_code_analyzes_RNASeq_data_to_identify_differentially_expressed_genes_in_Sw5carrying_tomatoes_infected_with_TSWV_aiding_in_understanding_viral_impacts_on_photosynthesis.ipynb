This notebook will analyze RNA-Seq data from Sw5-carrying tomatoes infected with TSWV.

```python
# Load necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load RNA-Seq data
rna_seq_data = pd.read_csv('rna_seq_data.csv')
```

The following analysis will identify differentially expressed genes related to photosynthesis.

```python
# Identify DEGs
DEGs = rna_seq_data[rna_seq_data['p_value'] < 0.05]
# Visualize DEGs
sns.barplot(x='gene', y='expression_change', data=DEGs)
```
