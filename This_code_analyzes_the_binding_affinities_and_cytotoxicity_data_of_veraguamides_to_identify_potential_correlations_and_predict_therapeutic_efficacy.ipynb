### Analysis of Veraguamide Data
This notebook analyzes the cytotoxicity and binding affinity data of veraguamides to predict their therapeutic potential.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for analysis
data = {'Compound': ['Veraguamide O', 'Veraguamide C'],
        'IC50_MDA_MB_231': [13, None],
        'IC50_MCF_7': [100, 0.32],
        'Binding_Affinity_sigma2': [720, 316]}
df = pd.DataFrame(data)

# Visualizing the data
plt.figure(figsize=(10, 5))
plt.bar(df['Compound'], df['IC50_MDA_MB_231'], color='blue', label='IC50 (MDA-MB-231)')
plt.bar(df['Compound'], df['IC50_MCF_7'], color='red', label='IC50 (MCF-7)', alpha=0.5)
plt.title('IC50 Values of Veraguamides')
plt.ylabel('IC50 (µM)')
plt.legend()
plt.show()
```
