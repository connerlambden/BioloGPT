### Analysis of Thyme Essential Oils
This notebook will analyze the chemical composition and bioactivity of essential oils from five Moroccan thyme species. We will visualize the data to identify trends and correlations.

```python
import pandas as pd
import plotly.express as px

# Data for analysis
data = {
    'Species': ['Thymus willdenowii', 'Thymus zygis subsp. gracilis', 'Thymus broussonetii', 'Thymus maroccanus', 'Thymus satureioides'],
    'Thymol Content (%)': [48.3, 41.5, 17.4, 38.1, 13.8],
    'DPPH IC50 (µg/mL)': [51.3, 70.6, 92.1, 114.8, 146.7],
    'Insecticidal LC50 (µL/mL)': [6.2, 7.1, 10.9, 7.2, 15.9]
}

df = pd.DataFrame(data)

# Create a scatter plot
fig = px.scatter(df, x='Thymol Content (%)', y='DPPH IC50 (µg/mL)',
                 size='Insecticidal LC50 (µL/mL)', hover_name='Species',
                 title='Thyme Essential Oils: Composition vs. Bioactivity')
fig.show()
```
