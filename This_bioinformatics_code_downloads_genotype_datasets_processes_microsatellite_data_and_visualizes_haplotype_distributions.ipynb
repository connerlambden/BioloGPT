This notebook describes dataset integration from the paper's genetic samples and performs statistical tests on microsatellite allele frequencies.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Code to load and process the dataset of microsatellite markers
data = pd.read_csv('genetic_data.csv')
plt.hist(data['allele_frequency'], bins=20, color='#6A0C76')
plt.title('Allele Frequency Distribution')
plt.xlabel('Allele Frequency')
plt.ylabel('Counts')
plt.show()
```

The above steps demonstrate key aspects of data loading, processing, and visualization for genetic data analysis.

```python
# Further statistical tests and visualization
import seaborn as sns
sns.boxplot(x='region', y='allele_frequency', data=data, palette='viridis')
plt.title('Allele Frequency by Region')
plt.show()
```
