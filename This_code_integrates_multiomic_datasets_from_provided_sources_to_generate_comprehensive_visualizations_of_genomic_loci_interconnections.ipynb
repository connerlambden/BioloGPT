We describe the extraction of genomic and imaging datasets to merge high-resolution histological data with GWAS outputs for cranial morphology studies.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Load real datasets from the provided sources
# Placeholder: replace with dataset loading code

genomic_data = pd.read_csv('path_to_cranial_GWAS_data.csv')
imaging_data = pd.read_csv('path_to_3D_histology_data.csv')

# Merge datasets on common identifiers
merged_data = pd.merge(genomic_data, imaging_data, on='sample_id')

# Plot sample relationship
plt.figure(figsize=(8,6))
plt.scatter(merged_data['genomic_marker'], merged_data['imaging_index'], color='#6A0C76')
plt.xlabel('Genomic Marker')
plt.ylabel('Imaging Index')
plt.title('Relationship between Genomic Markers and Imaging Indices')
plt.show()
```

This step-by-step notebook demonstrates merging genomic GWAS outputs with high-resolution imaging metrics, providing a framework for integrated data visualization.

```python
# Further analysis: computing correlation between the datasets
correlation = merged_data['genomic_marker'].corr(merged_data['imaging_index'])
print('Correlation between genomic markers and imaging indices:', correlation)
```
