Download and process genomic datasets for Kalanchoe species using the provided NCBI accession PRJNA944709.

```python
import requests
url = 'https://www.ncbi.nlm.nih.gov/bioproject/PRJNA944709'
response = requests.get(url)
if response.status_code == 200:
    print('Dataset downloaded and ready for analysis')
else:
    print('Error in downloading dataset')
```

Utilize MAFFT for multiple sequence alignment of BUSCO gene sets to construct a phylogenetic tree.

```python
import subprocess
command = ['mafft', '--auto', 'input_sequences.fasta']
subprocess.run(command, check=True)
print('Phylogenetic alignment completed')
```
