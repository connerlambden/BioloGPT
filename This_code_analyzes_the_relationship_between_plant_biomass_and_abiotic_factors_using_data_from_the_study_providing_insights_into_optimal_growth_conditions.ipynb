### Data Analysis of Grassland Biomass Relationships
This notebook will analyze the relationships between aboveground plant biomass and abiotic factors using the dataset from the study.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = 'https://doi.org/10.5281/zenodo.14509903'
data = pd.read_csv(url)

# Analyze the relationship between MAP and APB
plt.scatter(data['MAP'], data['APB'])
plt.title('Relationship between MAP and Aboveground Plant Biomass')
plt.xlabel('Mean Annual Precipitation (mm)')
plt.ylabel('Aboveground Plant Biomass (g/m2)')
plt.show()
```

This analysis will help visualize how changes in mean annual precipitation affect aboveground plant biomass in grasslands.
