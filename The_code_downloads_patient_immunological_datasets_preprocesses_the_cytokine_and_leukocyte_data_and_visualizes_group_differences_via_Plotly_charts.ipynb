This notebook details the extraction and visualization of leukocyte and immunological marker data from scabies patients compared to controls.

```python
import pandas as pd
import plotly.express as px

data = {'Parameter': ['Total WBCs', 'Lymphocytes', 'Monocytes', 'Neutrophils', 'Eosinophils', 'Basophils'],
        'Cases': [9.414, 34.248, 5.036, 52.586, 8.668, 0.908],
        'Controls': [7.900, 34.933, 5.006, 52.753, 1.680, 0.633]}
df = pd.DataFrame(data)

fig = px.bar(df, x='Parameter', y=['Cases', 'Controls'], barmode='group', title='Leukocyte Counts Comparison')
fig.show()
```
