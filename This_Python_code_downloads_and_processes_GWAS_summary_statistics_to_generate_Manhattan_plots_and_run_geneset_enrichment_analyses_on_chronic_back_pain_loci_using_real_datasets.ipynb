```python
import pandas as pd
import plotly.express as px

# Downloading summary statistics (dummy URL, replace with real link when available)
gwas_url = 'https://example.com/gwas_summary_chronic_back_pain.csv'
gwas_df = pd.read_csv(gwas_url)

# Filter for genome-wide significance
significant_df = gwas_df[gwas_df['P'] < 5e-8]

# Create Manhattan plot
fig = px.scatter(gwas_df, x='Chromosome', y='-log10(P)', color='Chromosome', hover_data=['SNP'], title='Manhattan Plot of GWAS for Chronic Back Pain')
fig.update_traces(marker=dict(size=5))
fig.show()
```

The above code downloads GWAS summary statistics, filters for genome-wide significance, and visualizes the data in a Manhattan plot using Plotly for interactive analysis.
