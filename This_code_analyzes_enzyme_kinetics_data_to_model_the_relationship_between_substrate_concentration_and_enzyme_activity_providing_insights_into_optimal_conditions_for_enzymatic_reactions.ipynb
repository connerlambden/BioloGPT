### Enzyme Kinetics Analysis
This notebook analyzes the relationship between substrate concentration and enzyme activity using Michaelis-Menten kinetics.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define substrate concentrations and corresponding enzyme activities
substrate_concentration = np.array([2, 4, 6, 8, 10])  # in %
enzyme_activity = np.array([20, 40, 70, 90, 95])  # hypothetical activity values

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(substrate_concentration, enzyme_activity, marker='o')
plt.title('Enzyme Activity vs. Substrate Concentration')
plt.xlabel('Substrate Concentration (%)')
plt.ylabel('Enzyme Activity (%)')
plt.grid(True)
plt.show()
```

### Analysis
This graph illustrates the relationship between substrate concentration and enzyme activity, highlighting the increase in activity with substrate concentration.
