## Correlation Analysis between Species Diversity and Carbon Sequestration

This notebook analyzes the relationship between species diversity indices and carbon sequestration rates using the dataset from the study.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

data = pd.DataFrame({
    'Species_Richness': [271],
    'Tree_Diversity_Index': [0.58],
    'Shrub_Diversity_Index': [0.81],
    'Carbon_Sequestration': [7.48]
})
```

### Step 1: Load and Inspect Data

```python
data.head()
```

### Step 2: Compute Pearson Correlation

```python
corr, _ = pearsonr(data['Species_Richness'], data['Carbon_Sequestration'])
print(f'Pearson Correlation: {corr}')
```

### Step 3: Visualize Correlation

```python
sns.scatterplot(x='Species_Richness', y='Carbon_Sequestration', data=data)
plt.title('Correlation between Species Richness and Carbon Sequestration')
plt.show()
```

### Step 4: Interpretation
The Pearson correlation coefficient indicates the strength and direction of the relationship between species richness and carbon sequestration.
