Load the snRNA-seq dataset from CNGBdb (accession CNP0004961) and preprocess the data to isolate oligodendrocyte lineage cells.

```python
import scanpy as sc
adata = sc.read_h5ad('path_to_data.h5ad')
# Filter for oligodendrocyte markers
oligodendrocytes = adata[adata.obs['cell_type'] == 'oligodendrocyte']
sc.pp.normalize_total(oligodendrocytes)
sc.pp.log1p(oligodendrocytes)
sc.tl.pca(oligodendrocytes)
sc.pl.pca(oligodendrocytes)
```

Perform differential expression analysis to compare SGK1 levels between diabetic and control samples within oligodendrocytes.

```python
sc.tl.rank_genes_groups(oligodendrocytes, 'condition', method='wilcoxon')
sc.pl.rank_genes_groups(oligodendrocytes, n_genes=20)
```

Generate a plotly graph to visualize the expression of SGK1 across OL subtypes related to diabetic status.

```python
import plotly.express as px
fig = px.violin(oligodendrocytes.obs, y='SGK1_expression', color='condition', box=True, points='all')
fig.show()
```
