# Codon Usage Analysis for Arginine
This notebook analyzes the codon usage for arginine across various species.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for codon usage
codon_data = {'Species': ['Human', 'Mouse', 'Fruit Fly'], 'CGU': [10, 12, 8], 'CGC': [15, 14, 10], 'CGA': [5, 6, 4], 'CGG': [7, 8, 5], 'AGA': [20, 18, 15], 'AGG': [3, 2, 1]}
df = pd.DataFrame(codon_data)

# Plotting the codon usage
plt.figure(figsize=(10, 6))
for codon in df.columns[1:]:
    plt.plot(df['Species'], df[codon], marker='o', label=codon)
plt.title('Codon Usage for Arginine Across Species')
plt.xlabel('Species')
plt.ylabel('Usage Frequency')
plt.legend()
plt.show()
```

This analysis provides insights into how different species utilize codons for arginine, which can reflect evolutionary adaptations.
