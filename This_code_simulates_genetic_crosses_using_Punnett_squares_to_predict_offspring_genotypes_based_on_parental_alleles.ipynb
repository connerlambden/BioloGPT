This notebook simulates genetic crosses using Punnett squares.

```python
import numpy as np
import pandas as pd

def punnett_square(parent1, parent2):
    # Generate all combinations of alleles
    offspring = [a+b for a in parent1 for b in parent2]
    return pd.Series(offspring).value_counts()

# Example usage
punnett_square('Aa', 'Aa')
```

This function calculates the expected genotype frequencies from a Punnett square.

```python
# Example of a Punnett square for a monohybrid cross
parent1 = 'Aa'
parent2 = 'Aa'
result = punnett_square(parent1, parent2)
result
```

The output shows the expected genotype frequencies for the offspring.
