### Step 1: Import Required Libraries
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

````python
# Load sequencing data
sequencing_data = pd.read_csv('sequencing_data.csv')

# Filter low-quality reads
filtered_data = sequencing_data[sequencing_data['quality'] >= 20]
````

### Step 2: Analyze Assembly Quality
```python
# Analyze assembly quality metrics
assembly_quality = filtered_data.groupby('assembly_id').agg({'contig_length': 'mean', 'n50': 'mean'})
assembly_quality.plot(kind='bar')
plt.title('Assembly Quality Metrics')
plt.xlabel('Assembly ID')
plt.ylabel('Quality Metrics')
plt.show()
```

````python
# Save the filtered data for further analysis
filtered_data.to_csv('filtered_sequencing_data.csv', index=False)
````
