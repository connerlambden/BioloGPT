This notebook step-by-step processes kinetic data and CYP1A1 gene expression from mouse models.

```python
import pandas as pd
import plotly.express as px
# Load datasets
kinetic_df = pd.read_csv('kinetic_data.csv')
gene_expr_df = pd.read_csv('CYP1A1_expression.csv')
# Merge datasets on sample_id
df = pd.merge(kinetic_df, gene_expr_df, on='sample_id')
# Scatter plot visualization
fig = px.scatter(df, x='half_life', y='expression_level', color='treatment', title='Half-life vs CYP1A1 Expression')
fig.show()
```

The code visualizes the relationship between drug half-life and CYP1A1 expression, providing insights into prodrug activation efficiency.

```python
import statsmodels.api as sm
X = df['expression_level']
Y = df['half_life']
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
print(model.summary())
```
