This notebook will analyze patient data to evaluate the long-term effects of SGLT2 inhibitors on graft survival.

```python
import pandas as pd

# Load patient data
patient_data = pd.read_csv('patient_data.csv')

# Analyze the impact of SGLT2 inhibitors on graft survival
results = patient_data.groupby('SGLT2i_use')['graft_survival'].mean()
results
```

The analysis shows the average graft survival rates for patients using SGLT2 inhibitors compared to those who are not.

```python
results.plot(kind='bar', title='Graft Survival Rates by SGLT2i Use')
```
