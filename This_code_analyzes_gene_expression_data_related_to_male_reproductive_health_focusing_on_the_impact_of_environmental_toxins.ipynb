### Gene Expression Analysis of Male Reproductive Health

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load gene expression data
data = pd.read_csv('gene_expression_data.csv')

# Analyze the impact of BPA on gene expression
bpa_effects = data[data['treatment'] == 'BPA']

# Visualize the results
plt.figure(figsize=(10,6))
sns.boxplot(x='gene', y='expression', data=bpa_effects)
plt.title('Impact of BPA on Gene Expression in Male Reproductive Organs')
plt.show()
```

### Discussion
This analysis provides insights into how environmental factors like BPA can alter gene expression related to male reproductive health.

```python
# Save the results for further analysis
bpa_effects.to_csv('bpa_gene_expression_analysis.csv', index=False)
```
