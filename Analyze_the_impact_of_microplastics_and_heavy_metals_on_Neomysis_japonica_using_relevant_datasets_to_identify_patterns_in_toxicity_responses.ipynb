This notebook will analyze the toxicity data of Neomysis japonica in response to microplastics and heavy metals.

```python
import pandas as pd

# Load relevant datasets
dataset_microplastics = pd.read_csv('microplastics_data.csv')
dataset_heavy_metals = pd.read_csv('heavy_metals_data.csv')

# Analyze the datasets
# Example analysis: calculate mortality rates
mortality_rates_microplastics = dataset_microplastics['mortality'].mean()
mortality_rates_heavy_metals = dataset_heavy_metals['mortality'].mean()

print(f'Mortality rate from microplastics: {mortality_rates_microplastics}')
print(f'Mortality rate from heavy metals: {mortality_rates_heavy_metals}')
```
