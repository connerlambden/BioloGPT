This notebook will analyze RNA-seq data to identify genes upregulated during telophase.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load RNA-seq data
data = pd.read_csv('telophase_rna_seq.csv')

# Analyze gene expression levels
gene_expression = data.groupby('gene').mean()

# Visualize results
plt.figure(figsize=(10,6))
sns.barplot(x=gene_expression.index, y=gene_expression.values)
plt.title('Gene Expression Levels During Telophase')
plt.xlabel('Genes')
plt.ylabel('Expression Level')
plt.xticks(rotation=90)
plt.show()
```

This analysis provides insights into the key genes involved in telophase.
