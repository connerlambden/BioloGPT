### Evolutionary Analysis of the tth Gene in Diptera
This notebook analyzes the conservation and divergence of the toothrin (tth) gene across various Diptera species to understand its evolutionary trajectory.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from Bio import Phylo
from io import StringIO

# Load gene sequence conservation data
# Assuming data is available in a CSV file with columns: species, sequence_identity
conservation_data = pd.read_csv('tth_conservation_diptera.csv')
```

### Sequence Conservation Across Species
Visualize the sequence identity of the tth gene across different Diptera species.

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='species', y='sequence_identity', data=conservation_data, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Diptera Species')
plt.ylabel('Sequence Identity (%)')
plt.title('Sequence Conservation of tth Gene Across Diptera')
plt.tight_layout()
plt.show()
```

### Phylogenetic Tree of tth Gene
Construct a phylogenetic tree to illustrate the evolutionary relationships based on tth gene sequences.

```python
# Load phylogenetic tree data in Newick format
newick_data = """
((SpeciesA:0.1, SpeciesB:0.2):0.3, SpeciesC:0.4);
"""
tree = Phylo.read(StringIO(newick_data), 'newick')
Phylo.draw(tree)
plt.show()
```

### Conclusion
The analysis demonstrates varying degrees of sequence conservation in the tth gene among Diptera species, suggesting differential evolutionary pressures and potential functional diversification.
