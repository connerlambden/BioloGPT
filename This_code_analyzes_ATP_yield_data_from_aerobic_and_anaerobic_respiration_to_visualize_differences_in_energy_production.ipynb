# ATP Yield Analysis of Aerobic vs Anaerobic Respiration
This notebook compares the ATP yield from aerobic and anaerobic respiration.

```python
import matplotlib.pyplot as plt
import numpy as np

types = ['Aerobic', 'Anaerobic']
yield = [36, 2]

plt.bar(types, yield, color=['blue', 'red'])
plt.title('ATP Yield in Aerobic vs Anaerobic Respiration')
plt.ylabel('ATP Yield')
plt.show()
```

The bar graph above illustrates the significant difference in ATP yield between aerobic and anaerobic respiration.
