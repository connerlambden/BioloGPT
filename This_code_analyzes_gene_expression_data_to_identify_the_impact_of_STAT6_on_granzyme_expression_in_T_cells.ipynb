### Gene Expression Analysis of Granzyme Regulation by STAT6

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('granzyme_expression_data.csv')
sns.boxplot(x='STAT6_activity', y='granzyme_expression', data=data)
plt.title('Granzyme Expression vs STAT6 Activity')
plt.show()
```

This analysis visualizes the relationship between STAT6 activity and granzyme expression levels in donor T cells.
