### Data Analysis of Femoral Fractures
This notebook analyzes the incidence of femoral fractures using GBD data.

```python
import pandas as pd

data = {'Year': [1990, 2021], 'Global Incidence Rate': [160.49, 146.57], 'Male Incidence Rate': [183.79, 155.53], 'Female Incidence Rate': [136.85, 137.55]}
df = pd.DataFrame(data)

# Calculate percentage change
percentage_change = ((df['Global Incidence Rate'].iloc[1] - df['Global Incidence Rate'].iloc[0]) / df['Global Incidence Rate'].iloc[0]) * 100
print(f'Percentage change in global incidence rate: {percentage_change:.2f}%')
```

### Summary of Findings
The analysis shows a decrease in the global incidence rate of femoral fractures from 1990 to 2021.

```python
# Visualize the data
import matplotlib.pyplot as plt

plt.bar(df['Year'], df['Global Incidence Rate'], color='blue', label='Global Incidence Rate')
plt.bar(df['Year'], df['Male Incidence Rate'], color='green', label='Male Incidence Rate')
plt.bar(df['Year'], df['Female Incidence Rate'], color='red', label='Female Incidence Rate')
plt.title('Incidence Rates of Femoral Fractures')
plt.xlabel('Year')
plt.ylabel('Incidence Rate per 100,000')
plt.legend()
plt.show()
```
