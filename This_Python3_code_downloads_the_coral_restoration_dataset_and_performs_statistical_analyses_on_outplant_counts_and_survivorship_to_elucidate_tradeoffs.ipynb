The code loads the coral restoration dataset from the provided sources, computes summary statistics, and generates graphs to compare survivorship across species.

```python
import pandas as pd
import plotly.express as px

# Load dataset (assuming CSV format from provided data)
df = pd.read_csv('coral_restoration_data.csv')

# Compute aggregate statistics for each species
summary = df.groupby('species').agg({'outplanted_count': 'sum', 'survivorship': 'mean'}).reset_index()
print(summary)

# Plotting using Plotly
fig = px.scatter(summary, x='species', y='outplanted_count', size='outplanted_count', color='survivorship', 
                 title='Outplant Counts and Average Survivorship by Species', 
                 labels={'outplanted_count':'Outplanted Count','species':'Coral Species','survivorship':'Avg Survivorship'})
fig.show()
```
