### Data Preparation
Load social network data and fitness metrics for analysis.

```python
import pandas as pd

# Load data
data = pd.read_csv('social_network_data.csv')

# Display the first few rows
print(data.head())
```

### Model Fitting
Fit a multilevel selection model to the data.

```python
import statsmodels.api as sm

# Fit a multilevel model
model = sm.MixedLM.from_formula('fitness ~ individual_trait + group_trait', data, groups=data['group_id'])
result = model.fit()
print(result.summary())
```

### Results Visualization
Visualize the selection gradients.

```python
import matplotlib.pyplot as plt

# Plotting selection gradients
plt.figure(figsize=(10, 6))
plt.bar(['Individual Trait', 'Group Trait'], [result.params['individual_trait'], result.params['group_trait']])
plt.title('Selection Gradients')
plt.ylabel('Gradient Value')
plt.show()
```
