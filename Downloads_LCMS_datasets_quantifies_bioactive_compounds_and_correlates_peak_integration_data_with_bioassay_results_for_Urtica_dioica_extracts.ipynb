This notebook loads LC-MS data from Urtica dioica extracts, performs peak integration, and plots the spectrum to quantify phenolic compounds.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('urtica_lc_ms_data.csv')
plt.figure(figsize=(10,5))
plt.plot(data['RT'], data['Intensity'], label='LC-MS Spectrum')
plt.xlabel('Retention Time (min)')
plt.ylabel('Intensity')
plt.title('LC-MS Profile of Urtica dioica Extract')
plt.legend()
plt.show()
peaks = data.groupby('Compound')['Intensity'].sum()
print(peaks)
```
