### Analysis of Biomass Reduction in Pasture Species
This notebook analyzes the effectiveness of different pasture species in reducing the biomass of annual riceflower. The data includes biomass reduction percentages at varying planting densities.

```python
import pandas as pd
import plotly.express as px

# Data for biomass reduction
data = {
    'Species': ['Premier digit grass', 'Rhodes grass', 'Sabi grass', 'Buffel grass'],
    'Biomass Reduction (%)': [49.4, 82.5, 73.7, 36.8]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot
fig = px.bar(df, x='Species', y='Biomass Reduction (%)', title='Biomass Reduction of Annual Riceflower by Pasture Species')
fig.show()
```
