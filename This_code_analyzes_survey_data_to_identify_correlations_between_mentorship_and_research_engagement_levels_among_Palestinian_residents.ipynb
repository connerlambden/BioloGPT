### Data Analysis of Resident Research Engagement
This notebook analyzes survey data from Palestinian resident doctors to explore the impact of mentorship on research engagement.

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('palestinian_residents_survey.csv')

# Analyze the correlation between mentorship and research engagement
correlation = data['mentorship'].corr(data['research_engagement'])
print(f'Correlation between mentorship and research engagement: {correlation}')
```

### Results
The analysis will provide insights into how mentorship influences research engagement among residents.

```python
# Further analysis can be conducted to visualize the data
import matplotlib.pyplot as plt

plt.scatter(data['mentorship'], data['research_engagement'])
plt.title('Mentorship vs Research Engagement')
plt.xlabel('Mentorship Level')
plt.ylabel('Research Engagement Level')
plt.show()
```
