### Soil Microbial Diversity Analysis
This notebook analyzes microbial diversity data from various soil samples to identify dominant phyla.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load microbial diversity data
data = {'Phyla': ['Actinobacteriota', 'Proteobacteria', 'Acidobacteriota', 'Firmicutes', 'Chloroflexota'], 'Relative Abundance': [42.72, 24.49, 15.00, 10.00, 8.79]}
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Phyla'], df['Relative Abundance'], color='skyblue')
plt.title('Relative Abundance of Soil Microbial Phyla')
plt.xlabel('Phyla')
plt.ylabel('Relative Abundance (%)')
plt.xticks(rotation=45)
plt.show()
```

### Discussion
This analysis provides insights into the dominant microbial communities in soil, which are crucial for understanding soil health and fertility.

```python
# Save the plot
plt.savefig('soil_microbial_diversity.png')
```
