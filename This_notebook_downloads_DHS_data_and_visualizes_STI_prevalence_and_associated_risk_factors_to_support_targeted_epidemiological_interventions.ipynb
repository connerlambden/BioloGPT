Begin by loading the DHS dataset and extracting relevant columns such as country, prevalence percentages, and risk factors.

```python
import pandas as pd
import plotly.express as px
# Load DHS data (assuming a CSV file is provided from DHS)
df = pd.read_csv('dhs_data.csv')
# Filter dataset for selected Southeast Asian countries
countries = ['Cambodia', 'Indonesia', 'Timor-Leste', 'Myanmar', 'Philippines']
df_filtered = df[df['Country'].isin(countries)]
fig = px.bar(df_filtered, x='Country', y='Prevalence', color='Country', title='STI Prevalence by Country')
fig.show()
```

The next cell explores associations between socio-demographic factors and STI prevalence using regression analysis.

```python
import statsmodels.api as sm
# Assume 'Age', 'Education', 'WealthIndex' are columns in the dataset
X = df_filtered[['Age', 'Education', 'WealthIndex']]
X = sm.add_constant(X)
y = df_filtered['Prevalence']
model = sm.OLS(y, X).fit()
print(model.summary())
```
