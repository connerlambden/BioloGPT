### Data Preparation
Load SNP and metabolite data for analysis.

```python
import pandas as pd

# Load SNP data
snp_data = pd.read_csv('snp_data.csv')
# Load metabolite data
metabolite_data = pd.read_csv('metabolite_data.csv')
```

### Data Analysis
Merge datasets and analyze correlations.

```python
# Merge datasets on species
merged_data = pd.merge(snp_data, metabolite_data, on='species')

# Calculate correlation
correlation = merged_data.corr()
```

### Results Visualization
Visualize the correlation matrix.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation between SNPs and Secondary Metabolites')
plt.show()
```
