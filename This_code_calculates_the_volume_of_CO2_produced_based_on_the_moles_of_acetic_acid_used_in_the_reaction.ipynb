# CO2 Production Calculation
This notebook calculates the volume of CO2 produced from a given amount of acetic acid.

```python
def calculate_co2_volume(moles_acetic_acid):
    # At STP, 1 mole of gas occupies 22.4 liters
    co2_moles = moles_acetic_acid  # 1:1 ratio from the reaction
    volume_co2 = co2_moles * 22.4  # in liters
    return volume_co2

# Example usage:
calculate_co2_volume(0.1)  # Replace 0.1 with the desired moles of acetic acid.
```

This function will return the volume of CO2 produced in liters based on the input moles of acetic acid.

```python
# Example output
volume = calculate_co2_volume(0.1)
print(f'Volume of CO2 produced: {volume} liters')
```
