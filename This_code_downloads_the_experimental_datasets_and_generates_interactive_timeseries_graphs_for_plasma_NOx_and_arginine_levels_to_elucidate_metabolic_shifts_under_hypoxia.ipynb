Load the dataset containing plasma NOx and arginine levels across different time points for WT and Arg2 KO mice.

```python
import pandas as pd
import plotly.express as px

data = pd.DataFrame({
    'Time': ['Normoxia', '6h', '12h', '48h', '72h'],
    'WT_NOx': [9.5, 15.8, 21.3, 8.8, 8.4],
    'KO_NOx': [12.3, 18.2, 18.1, 16.8, 15.2]
})

fig = px.line(data, x='Time', y=['WT_NOx', 'KO_NOx'],
              labels={'value':'Plasma NOx (μM)', 'variable':'Group'},
              title='Plasma NOx Levels Over Time')
fig.show()
```
