This notebook will analyze student performance data from collaborative and individual lab activities.

```python
import pandas as pd

# Load data
collab_data = pd.read_csv('collaborative_learning_data.csv')
indiv_data = pd.read_csv('individual_learning_data.csv')

# Analyze retention rates
collab_retention = collab_data['retention_rate'].mean()
indiv_retention = indiv_data['retention_rate'].mean()

print(f'Collaborative Retention Rate: {collab_retention}')
print(f'Individual Retention Rate: {indiv_retention}')
```

This analysis will help determine if collaborative learning significantly improves retention rates.

```python
# Visualization of retention rates
import matplotlib.pyplot as plt

labels = ['Collaborative', 'Individual']
retention_rates = [collab_retention, indiv_retention]

plt.bar(labels, retention_rates, color=['#6A0C76', '#FF5733'])
plt.ylabel('Retention Rate')
plt.title('Retention Rates: Collaborative vs Individual Learning')
plt.show()
```
