### Plasma Metabolite Analysis
This section analyzes the changes in plasma metabolite levels before and after a competitive match.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for metabolites
data = {
    'Metabolite': ['Histidine', 'Asparagine', 'Kynurenine', 'Phenylalanine', 'Glutamine', 'Proline', 'Lysine', 'Tryptophan', 'Tyrosine', 'Methionine', 'Alanine', 'Serine', 'Pyruvate'],
    'Post_Match': [1.4504, 0.3830, 0.10154, 0.5812, 0.5267, 0.4748, 1.783, 2090, 0.3278, 0.2780, 5.414, 1.565, 0.4689]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10,6))
plt.bar(df['Metabolite'], df['Post_Match'], color='skyblue')
plt.title('Plasma Metabolite Levels Post-Match')
plt.xlabel('Metabolites')
plt.ylabel('Post-Match Levels')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
