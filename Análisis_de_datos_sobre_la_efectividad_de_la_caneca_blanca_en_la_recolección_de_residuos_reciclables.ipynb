Análisis de datos de recolección de residuos en comunidades con canecas blancas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos de recolección
data = pd.read_csv('recoleccion_residuos.csv')

# Visualizar datos
plt.figure(figsize=(10,5))
plt.bar(data['Comunidad'], data['Residuos_Reciclados'])
plt.title('Residuos reciclados por comunidad')
plt.xlabel('Comunidad')
plt.ylabel('Cantidad de residuos reciclados')
plt.show()
```

Este análisis permite visualizar la efectividad de la caneca blanca en diferentes comunidades.
