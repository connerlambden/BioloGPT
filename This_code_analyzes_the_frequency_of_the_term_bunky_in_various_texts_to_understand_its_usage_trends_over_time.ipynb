# Analyzing the frequency of the term 'bunky' in literature over time.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data representing the frequency of 'bunky' over decades
data = {'Decade': ['1920s', '1930s', '1940s', '1950s', '1960s', '1970s', '1980s', '1990s', '2000s', '2010s'],
        'Frequency': [5, 3, 2, 1, 1, 0.5, 0.2, 0.1, 0.05, 0.01]}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10,5))
plt.plot(df['Decade'], df['Frequency'], marker='o')
plt.title('Historical Usage of the Term "Bunky"')
plt.xlabel('Decade')
plt.ylabel('Frequency in Literature')
plt.grid()
plt.show()
```
