### Analysis of Antimicrobial Scaffold Efficacy
This notebook analyzes the data on bacterial CFU reduction and BM-MSC viability from the study.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from the study
data = {
    'Condition': ['Control', 'Plain Scaffold', 'Antimicrobial Scaffold'],
    'S. aureus CFU': [211, None, 25],
    'E. coli CFU': [228, None, 6],
    'BM-MSC Viability Day 7 (%)': [None, 74, 97]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 5))
plt.bar(df['Condition'], df['S. aureus CFU'], label='S. aureus CFU')
plt.bar(df['Condition'], df['E. coli CFU'], label='E. coli CFU', bottom=df['S. aureus CFU'])
plt.title('Bacterial Colonies (CFU) Comparison')
plt.ylabel('CFU Count')
plt.legend()
plt.show()
```
