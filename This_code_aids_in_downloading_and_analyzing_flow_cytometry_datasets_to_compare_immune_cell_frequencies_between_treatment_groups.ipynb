Load and preprocess flow cytometry data from the study's supplementary files.

```python
import pandas as pd

data = pd.read_csv('flow_data.csv')
# Perform normalization and group by treatment
summary = data.groupby('treatment').mean()
print(summary)
```

Visualize differences in key immune populations using a bar graph.

```python
import plotly.express as px

fig = px.bar(summary, barmode='group', title='Immune Cell Populations across Treatments')
fig.show()
```

This analysis highlights differences between sensitive and resistant models, aiding in hypothesis testing.
