### Phage Titre Analysis
This notebook analyzes the effects of different buffer compositions on phage titres.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load phage titre data
phage_data = pd.read_csv('phage_titre_data.csv')

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(phage_data['Buffer'], phage_data['Titre'])
plt.xlabel('Buffer Composition')
plt.ylabel('Phage Titre (PFU/mL)')
plt.title('Phage Titre Comparison by Buffer Composition')
plt.show()
```
