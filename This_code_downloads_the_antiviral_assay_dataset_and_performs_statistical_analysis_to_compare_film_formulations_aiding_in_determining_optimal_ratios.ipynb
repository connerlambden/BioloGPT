Step-by-step analysis of film antiviral efficacy using raw PFU data from in vitro assays.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load dataset
 data = pd.read_csv('antiviral_assay_data.csv')
# Compute statistics
 summary = data.groupby('formulation').agg({'PFU_reduction': ['mean', 'std']})
print(summary)
# Plotting the results
plt.errorbar(summary.index, summary[('PFU_reduction', 'mean')], yerr=summary[('PFU_reduction', 'std')], fmt='o')
plt.title('Antiviral Efficacy by Formulation')
plt.xlabel('Formulation Ratio')
plt.ylabel('Mean PFU Reduction')
plt.show()
```
