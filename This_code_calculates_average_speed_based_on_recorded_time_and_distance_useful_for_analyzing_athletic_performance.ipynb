### Average Speed Calculation
This notebook calculates the average speed based on the distance of 400 meters and various recorded times.

```python
# Function to calculate average speed

def calculate_average_speed(distance, time):
    return distance / time

# Example usage
recorded_times = [80, 79.5, 78, 77.5]  # in seconds
results = {time: calculate_average_speed(400, time) for time in recorded_times}
results
```
