Análisis de diversidad genética en plantas.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('plant_data.csv')
sns.boxplot(x='species', y='genetic_diversity', data=data)
plt.title('Diversidad Genética por Especie')
plt.show()
```
