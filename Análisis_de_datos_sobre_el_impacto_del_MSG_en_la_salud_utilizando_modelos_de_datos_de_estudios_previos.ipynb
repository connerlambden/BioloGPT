Este análisis examina cómo el MSG afecta la salud a través de datos de estudios previos.

```python
import pandas as pd
# Cargar datos de estudios sobre MSG
dataset = pd.read_csv('data_studies_msg.csv')
# Análisis de correlaciones entre MSG y efectos en la salud.
```

Los resultados se analizarán para identificar patrones y correlaciones.

```python
results = dataset.corr()  # Calcular correlaciones.
```

Los resultados se visualizarán para facilitar la interpretación.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(results, annot=True)
plt.show()
```
