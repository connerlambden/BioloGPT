Load the ant community dataset and relevant metadata directly from the provided Zenodo link to reproduce community composition and trait analyses.

```python
import pandas as pd
import plotly.express as px
# Download dataset from the provided Zenodo link
url = 'https://zenodo.org/records/14792981'
df = pd.read_csv(url)
# Create a scatter plot of thermal preference vs. species richness
fig = px.scatter(df, x='thermal_preference', y='species_richness', color='habitat', title='Thermal Preference and Species Richness')
fig.show()
```

The above code visualizes the relationship between thermal preference and species richness in ant communities, highlighting urban versus rural differences.

```python
# Additional analyses can include group comparisons using statistical tests
import statsmodels.api as sm
model = sm.OLS(df['species_richness'], sm.add_constant(df[['thermal_preference']])).fit()
print(model.summary())
```

This completes the basic analysis, providing insights on how thermal traits influence community composition in urban settings.
