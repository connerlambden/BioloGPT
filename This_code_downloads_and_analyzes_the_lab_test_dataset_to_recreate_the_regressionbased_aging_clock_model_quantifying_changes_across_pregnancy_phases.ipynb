Retrieve and inspect the dataset used in the study to perform regression analysis correlating lab values with apparent age.

```python
import pandas as pd
import statsmodels.api as sm
# Download dataset from GitHub repository
url = 'https://github.com/AlonLabWIS/PregAging/raw/main/data.csv'
data = pd.read_csv(url)
# Example regression: Predict age from selected lab test values
X = data[['RBC', 'MCV', 'RDW', 'Hematocrit']]
X = sm.add_constant(X)
y = data['Age']
model = sm.OLS(y, X).fit()
print(model.summary())
```

The notebook performs a regression analysis to quantify the relationship between key lab tests and apparent age, replicating part of the study's findings.

```python
# Plot the predicted age vs actual age using Plotly
import plotly.express as px
predicted_age = model.predict(X)
data['PredictedAge'] = predicted_age
fig = px.scatter(data, x='Age', y='PredictedAge', title='Apparent Age Prediction')
fig.show()
```
