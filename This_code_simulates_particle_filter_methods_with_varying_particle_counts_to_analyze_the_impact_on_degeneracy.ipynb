This notebook explores the effects of particle count on the performance of SMC methods.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate likelihoods for different particle counts
particle_counts = [8, 32, 128, 512, 2048, 8192]
likelihoods = np.random.rand(len(particle_counts)) * np.log(particle_counts)

plt.plot(particle_counts, likelihoods, marker='o')
plt.xscale('log')
plt.xlabel('Number of Particles')
plt.ylabel('Likelihood')
plt.title('Effect of Particle Count on Likelihood')
plt.grid(True)
plt.show()
```

The graph illustrates how increasing particle counts can affect the likelihood in SMC methods.
