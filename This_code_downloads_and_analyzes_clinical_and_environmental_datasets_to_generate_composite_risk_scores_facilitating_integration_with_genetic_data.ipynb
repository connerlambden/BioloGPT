In this notebook, we detail steps to download patient clinical data, compute polyenviromic risk and protective scores, and visualize inter-group differences.

```python
import pandas as pd
import plotly.express as px

data = pd.read_csv('clinical_environmental_data.csv')
# Compute risk scores and visualize
fig = px.bar(data, x='Diagnosis', y='Count', color='Group', barmode='group', title='Diagnostic Distribution')
fig.show()
```

This segment explains the data cleaning, normalization, and score aggregation steps in detail for reproducibility.

```python
data['PERS'] = (data['Risk_Factor_Present'] * data['Log_Odds_Ratio']).groupby(data['Participant_ID']).transform('sum') / data['Total_Factors']
fig2 = px.histogram(data, x='PERS', color='Group', nbins=30, title='Distribution of Polyenviromic Risk Scores')
fig2.show()
```
