### Growth Analysis of Serratia marcescens
This notebook analyzes the growth data of Serratia marcescens under different temperature and time conditions.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Temperature (°C)': [5, 15, 25, 35, 40], 'Growth (Relative Units)': [0, 1, 3, 5, 2]}
df = pd.DataFrame(data)
plt.bar(df['Temperature (°C)'], df['Growth (Relative Units)'])
plt.title('Growth of Serratia marcescens at Various Temperatures')
plt.xlabel('Temperature (°C)')
plt.ylabel('Growth (Relative Units)')
plt.show()
```
