### Plant Growth Analysis in Martian Conditions
This notebook analyzes the growth data of Swiss chard in various compost and regolith mixtures.

```python
import pandas as pd
import matplotlib.pyplot as plt

growth_data = {
    'C:S ratio': ['1:0', '3:1', '1:1', '1:3', '0:1'],
    'Fertilized Growth (g)': [991, 951, 821, 640, 431],
    'Non-Fertilized Growth (g)': [118, 114, 84, 38, 2]
}
df = pd.DataFrame(growth_data)

plt.bar(df['C:S ratio'], df['Fertilized Growth (g)'], label='Fertilized', alpha=0.6)
plt.bar(df['C:S ratio'], df['Non-Fertilized Growth (g)'], label='Non-Fertilized', alpha=0.6)
plt.title('Swiss Chard Growth in Compost and Mars Regolith Simulant')
plt.xlabel('C:S Ratio')
plt.ylabel('Growth (g)')
plt.legend()
plt.show()
```
