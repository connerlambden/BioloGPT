This notebook provides a step-by-step analysis of gene expression profiles from PDAC samples and correlates the expression of hERG1 and β1 integrin with patient survival data.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Code to download and process the dataset
url = 'https://www.ebi.ac.uk/biostudies/files/E-GEOD-18277/E-GEOD-18277.json'
df = pd.read_json(url)
df.head()
```

The following cell will create a scatter plot correlating the expression levels of hERG1 and β1 integrin to explore potential predictive markers for statin responsiveness.

```python
plt.figure(figsize=(10,6))
plt.scatter(df['hERG1_expression'], df['β1_integrin_expression'], c='blue', alpha=0.7)
plt.xlabel('hERG1 Expression')
plt.ylabel('β1 Integrin Expression')
plt.title('Correlation between hERG1 and β1 Integrin Expression in PDAC samples')
plt.show()
```
