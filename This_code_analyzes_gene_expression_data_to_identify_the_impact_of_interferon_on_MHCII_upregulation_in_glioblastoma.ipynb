### Data Analysis of Interferon Response in Glioblastoma

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('interferon_mhc_data.csv')
sns.boxplot(x='Treatment', y='MHCII_Expression', data=data)
plt.title('MHCII Expression Levels by Treatment')
plt.show()
```

This analysis visualizes the differences in MHCII expression levels between treated and untreated glioblastoma samples.

```python
# Example code to analyze gene expression
import numpy as np
import statsmodels.api as sm

model = sm.OLS(data['MHCII_Expression'], data[['IFN_Gamma', 'Treatment']]).fit()
model.summary()
```
