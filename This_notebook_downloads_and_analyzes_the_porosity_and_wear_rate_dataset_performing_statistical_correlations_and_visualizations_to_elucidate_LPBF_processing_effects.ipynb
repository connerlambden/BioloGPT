This notebook examines the relationship between processing parameters, porosity, and tribological performance in 316L stainless steel.

```python
import pandas as pd
import plotly.express as px

# Define dataset based on extracted data
data = pd.DataFrame([
    {'Sample': 'S1', 'Power': 120, 'Speed': 700, 'Porosity': 4.3},
    {'Sample': 'S2', 'Power': 120, 'Speed': 900, 'Porosity': 6.5},
    {'Sample': 'S3', 'Power': 120, 'Speed': 1100, 'Porosity': 9.1},
    {'Sample': 'S4', 'Power': 150, 'Speed': 700, 'Porosity': 3.4},
    {'Sample': 'S5', 'Power': 150, 'Speed': 900, 'Porosity': 5.3},
    {'Sample': 'S6', 'Power': 150, 'Speed': 1100, 'Porosity': 7.1},
    {'Sample': 'S7', 'Power': 180, 'Speed': 700, 'Porosity': 1.7},
    {'Sample': 'S8', 'Power': 180, 'Speed': 900, 'Porosity': 3.2},
    {'Sample': 'S9', 'Power': 180, 'Speed': 1100, 'Porosity': 5.4}
])

# Create scatter plot
fig = px.scatter(data, x='Speed', y='Porosity', color='Power', text='Sample', title='Porosity vs Scanning Speed')
fig.show()
```
