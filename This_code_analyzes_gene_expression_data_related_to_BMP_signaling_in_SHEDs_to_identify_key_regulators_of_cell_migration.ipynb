This notebook will analyze gene expression data from SHEDs cultured with hPL, focusing on BMP signaling pathways.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('sheds_bmp_signaling.csv')
sns.boxplot(x='BMP_Signaling', y='Cell_Migration', data=data)
plt.title('Effect of BMP Signaling on SHED Migration')
plt.show()
```
