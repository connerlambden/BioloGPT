### Analyzing Crystal Packing Interactions
This notebook will explore the relationships between amino acid sequences and their crystallization propensity.

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load dataset of protein structures
protein_data = pd.read_csv('protein_crystal_data.csv')

# Analyze packing interactions
packing_analysis = protein_data.groupby('amino_acid').agg({'crystallization_success': 'mean'})
packing_analysis.sort_values(by='crystallization_success', ascending=False).head(10)
```

### Results
The analysis reveals which amino acids are most associated with successful crystallization.
