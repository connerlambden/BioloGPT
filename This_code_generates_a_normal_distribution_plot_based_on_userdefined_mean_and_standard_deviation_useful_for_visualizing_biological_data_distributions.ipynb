This notebook allows users to input their own mean and standard deviation to visualize a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# User-defined parameters
mean = 0
std_dev = 1

# Generate data
x = np.linspace(-5, 5, 100)
y = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# Plotting
plt.plot(x, y)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid()
plt.show()
```
