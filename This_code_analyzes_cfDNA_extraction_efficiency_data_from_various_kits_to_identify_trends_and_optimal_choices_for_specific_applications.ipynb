# cfDNA Extraction Kit Analysis
This notebook analyzes the efficiency of various cfDNA extraction kits based on recovery rates and costs.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for extraction kits
kits_data = {
    'Kit': ['QIAamp', 'MagMAX', 'VAHTS', 'JBS'],
    'Recovery Efficiency (%)': [90, 75, 90, 85],
    'Cost': ['High', 'Moderate', 'Moderate', 'Low']
}
df = pd.DataFrame(kits_data)

# Plotting the recovery efficiency
plt.bar(df['Kit'], df['Recovery Efficiency (%)'], color='blue')
plt.title('cfDNA Extraction Kit Recovery Efficiency')
plt.xlabel('Kit Name')
plt.ylabel('Recovery Efficiency (%)')
plt.show()
```

## Discussion
This analysis provides insights into the performance of different cfDNA extraction kits, aiding in the selection process for specific research needs.
