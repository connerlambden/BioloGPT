### RNA-seq Analysis of DGKz Deficient B Cells
This analysis aims to identify differentially expressed genes in DGKz-deficient B cells compared to wild-type controls.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load RNA-seq data
wt_data = pd.read_csv('wild_type_b_cells.csv')
dgkz_data = pd.read_csv('dgkz_deficient_b_cells.csv')

# Perform differential expression analysis
results = stats.ttest_ind(wt_data['expression'], dgkz_data['expression'])

# Compile results
results_df = pd.DataFrame({'gene': wt_data['gene'], 'p_value': results.pvalue})
results_df['significant'] = results_df['p_value'] < 0.05
results_df.to_csv('differential_expression_results.csv', index=False)
```
