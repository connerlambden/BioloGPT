### Population Dynamics Simulation in Fragmented Landscapes
This notebook simulates the impact of home-range sizes on population dynamics in fragmented landscapes.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
home_range_sizes = [1, 5, 10]  # Example home range sizes
fragmentation_levels = np.linspace(0, 1, 100)  # 0 to 100% fragmentation
carrying_capacity = []

# Simulate carrying capacity based on home range size and fragmentation
for size in home_range_sizes:
    capacity = 100 * (1 - fragmentation_levels) ** (size / 10)
    carrying_capacity.append(capacity)

# Plot results
plt.figure(figsize=(10, 6))
for i, size in enumerate(home_range_sizes):
    plt.plot(fragmentation_levels * 100, carrying_capacity[i], label=f'Home Range Size: {size}')
plt.title('Carrying Capacity vs. Fragmentation Level')
plt.xlabel('Fragmentation Level (%)')
plt.ylabel('Carrying Capacity (%)')
plt.legend()
plt.grid()
plt.show()
```

### Discussion
This simulation illustrates how carrying capacity declines with increasing fragmentation, particularly for species with larger home ranges.
