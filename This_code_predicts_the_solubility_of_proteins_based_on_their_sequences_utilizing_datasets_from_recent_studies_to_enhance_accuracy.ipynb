### Protein Solubility Prediction Notebook
This notebook utilizes the NetSolP model to predict the solubility of various proteins based on their sequences.

```python
# Import necessary libraries
import pandas as pd
import requests

# Load protein sequences from a dataset
sequences = pd.read_csv('protein_sequences.csv')

# Function to predict solubility using NetSolP
def predict_solubility(sequence):
    response = requests.post('https://services.healthtech.dtu.dk/service.php?NetSolP-1.0', data={'sequence': sequence})
    return response.json()
```

### Predictions
Run the solubility predictions for the loaded sequences.

```python
# Apply the prediction function to each sequence
sequences['solubility'] = sequences['sequence'].apply(predict_solubility)
```
