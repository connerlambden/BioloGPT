This notebook will analyze RNA-seq data from Ku70-deficient mice to identify compensatory DNA repair pathways.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load RNA-seq data
data = pd.read_csv('ku70_deficient_rna_seq.csv')

# Analyze expression levels of DNA repair genes
repair_genes = ['Ku70', 'Ku80', 'DNA-PKcs', 'Rad51', 'XRCC1']
filtered_data = data[data['gene'].isin(repair_genes)]

# Visualize expression levels
plt.figure(figsize=(10,6))
sns.barplot(x='gene', y='expression_level', data=filtered_data)
plt.title('Expression Levels of DNA Repair Genes in Ku70-Deficient Mice')
plt.show()
```
