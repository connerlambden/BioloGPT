### Analysis of Rosa damascena Essential Oil Effects
This notebook analyzes the cytotoxic and genotoxic effects of Rosa damascena essential oil based on existing research data.

```python
import pandas as pd

# Load relevant datasets
dataset = pd.read_csv('rosa_damascena_effects.csv')

# Analyze the effects based on concentration
results = dataset.groupby('concentration').agg({'genotoxicity': 'mean', 'cytotoxicity': 'mean'})
results
```

### Results
The results show the average genotoxicity and cytotoxicity at various concentrations of Rosa damascena essential oil.

```python
results.plot(kind='bar', title='Cytotoxic and Genotoxic Effects of Rosa damascena Essential Oil')
```
