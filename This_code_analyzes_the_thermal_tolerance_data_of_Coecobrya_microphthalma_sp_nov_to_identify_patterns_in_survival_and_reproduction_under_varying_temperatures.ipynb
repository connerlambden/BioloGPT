### Thermal Tolerance Analysis of Coecobrya microphthalma sp. nov.
This notebook analyzes the survival rates and reproductive capacity of Coecobrya microphthalma sp. nov. at different temperatures.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Temperature (°C)': [27, 30, 32],
    'Survival Rate (%)': [100, 57, 50],
    'Egg Laying Duration (days)': [6.55, 4.80, 3.50]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 5))
plt.bar(df['Temperature (°C)'], df['Survival Rate (%)'], color='blue', alpha=0.6, label='Survival Rate (%)')
plt.bar(df['Temperature (°C)'], df['Egg Laying Duration (days)'], color='orange', alpha=0.6, label='Egg Laying Duration (days)')
plt.title('Thermal Tolerance of Coecobrya microphthalma sp. nov.')
plt.xlabel('Temperature (°C)')
plt.ylabel('Rate / Duration')
plt.legend()
plt.show()
```

### Discussion
This analysis highlights the significant impact of temperature on the survival and reproductive capacity of Coecobrya microphthalma sp. nov.
