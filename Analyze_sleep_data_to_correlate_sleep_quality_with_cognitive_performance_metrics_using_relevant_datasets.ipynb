This notebook will analyze sleep quality data and its impact on cognitive performance using datasets from recent studies.

```python
import pandas as pd

# Load sleep quality and cognitive performance datasets
data_sleep = pd.read_csv('sleep_quality_data.csv')
data_cognitive = pd.read_csv('cognitive_performance_data.csv')

# Merge datasets on participant ID
merged_data = pd.merge(data_sleep, data_cognitive, on='participant_id')

# Analyze correlation between sleep quality and cognitive scores
correlation = merged_data['sleep_quality'].corr(merged_data['cognitive_score'])
print(f'Correlation between sleep quality and cognitive performance: {correlation}')
```

This analysis will provide insights into how sleep quality affects cognitive performance, potentially guiding future interventions.
