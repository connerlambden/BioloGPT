Analiza sekwencji DNA mitochondriów i chloroplastów.

```python
# Importowanie bibliotek
import pandas as pd
import matplotlib.pyplot as plt

# Wczytanie danych
mitochondria_data = pd.read_csv('mitochondria_sequences.csv')
chloroplast_data = pd.read_csv('chloroplast_sequences.csv')

# Analiza danych
# ... (analiza sekwencji) 

# Wizualizacja wyników
plt.plot(mitochondria_data['length'], label='Mitochondria')
plt.plot(chloroplast_data['length'], label='Chloroplasty')
plt.legend()
plt.show()
```

Wizualizacja długości sekwencji DNA mitochondriów i chloroplastów.

```python
# Wizualizacja długości sekwencji
plt.hist(mitochondria_data['length'], alpha=0.5, label='Mitochondria')
plt.hist(chloroplast_data['length'], alpha=0.5, label='Chloroplasty')
plt.legend()
plt.title('Długość sekwencji DNA')
plt.xlabel('Długość (bp)')
plt.ylabel('Częstość')
plt.show()
```
