### Data Preparation
Load genomic data for Bombella apis and phage exposure datasets.

```python
import pandas as pd

# Load datasets
df_crispr = pd.read_csv('crispr_data.csv')
df_phage = pd.read_csv('phage_data.csv')
```

### Data Analysis
Merge datasets and analyze correlations.

```python
# Merge datasets based on common identifiers
df_merged = pd.merge(df_crispr, df_phage, on='sample_id')

# Calculate correlation
correlation = df_merged.corr()['crispr_distribution']['phage_exposure']
```

### Results
Display the correlation result.

```python
print(f'Correlation between CRISPR distribution and phage exposure: {correlation}')
```
