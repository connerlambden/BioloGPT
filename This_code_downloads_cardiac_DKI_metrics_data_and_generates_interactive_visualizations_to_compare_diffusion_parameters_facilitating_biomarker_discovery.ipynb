This notebook retrieves the diffusion metrics from the cardiac DKI study and creates an interactive bar chart for visual comparison.

```python
import pandas as pd
import plotly.express as px
# Cardiac DKI metrics data
data = {'Metric': ['MD', 'FA', 'MK', 'AK', 'RK'], 'Value': [1.66, 0.31, 0.32, 0.27, 0.35]}
df = pd.DataFrame(data)
fig = px.bar(df, x='Metric', y='Value', title='Cardiac DKI Metrics')
fig.show()
```

The interactive visualization helps elucidate the differences among the metrics, which is critical for understanding myocardial microstructure.
