This notebook loads relevant datasets and produces a bar chart to visualize the projected percentage increase in chemical and pharmaceutical product exports under the EMFTA scenario.

```python
import plotly.graph_objects as go

# Data for the graph
categories = ['Chemical Products', 'Pharmaceutical Products']
percentage_increase = [35, 21]

fig = go.Figure([go.Bar(x=categories, y=percentage_increase, marker_color='#6A0C76')])

fig.update_layout(title='Projected Increase in Swiss Exports Under EMFTA',
                  xaxis_title='Product Category',
                  yaxis_title='Percentage Increase')

fig.show()
```
