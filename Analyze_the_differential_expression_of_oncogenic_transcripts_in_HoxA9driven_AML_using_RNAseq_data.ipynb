This notebook will analyze RNA-seq data to identify differentially expressed oncogenic transcripts in HoxA9-driven AML.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load RNA-seq data
data = pd.read_csv('path_to_rna_seq_data.csv')

# Perform differential expression analysis
results = stats.ttest_ind(data['HoxA9wt'], data['HoxA9AA'])

# Output results
results_df = pd.DataFrame({'Gene': data['Gene'], 'p-value': results.pvalue})
results_df.to_csv('differential_expression_results.csv', index=False)
```
