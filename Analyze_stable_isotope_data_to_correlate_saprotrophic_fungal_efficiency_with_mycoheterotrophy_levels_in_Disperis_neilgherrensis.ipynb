This notebook analyzes stable isotope data from *Disperis neilgherrensis* and its associated saprotrophic fungi.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
fungal_data = pd.read_csv('fungal_efficiency.csv')
myco_data = pd.read_csv('mycoheterotrophy.csv')

# Merge datasets
merged_data = pd.merge(fungal_data, myco_data, on='microhabitat')

# Plotting
plt.scatter(merged_data['fungal_efficiency'], merged_data['mycoheterotrophy'])
plt.xlabel('Fungal Efficiency')
plt.ylabel('Mycoheterotrophy Levels')
plt.title('Correlation between Fungal Efficiency and Mycoheterotrophy')
plt.show()
```
