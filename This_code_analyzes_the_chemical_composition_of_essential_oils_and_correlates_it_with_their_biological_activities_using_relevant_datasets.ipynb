### Analysis of Essential Oil Composition and Biological Activity
This notebook will analyze the composition of Pulicaria burchardii essential oil and its biological effects, focusing on the correlation between specific compounds and their antimicrobial/antiviral activities.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for essential oil composition
data = {'Compound': ['1-oxo-bisabolone', 'Eucalyptol', 'τ-cadinol'], 'Percentage': [65.09, 13.87, 4.21]}

df = pd.DataFrame(data)

# Plotting the composition
plt.bar(df['Compound'], df['Percentage'], color='green')
plt.title('Composition of Pulicaria burchardii Essential Oil')
plt.xlabel('Compounds')
plt.ylabel('Percentage (%)')
plt.show()
```
