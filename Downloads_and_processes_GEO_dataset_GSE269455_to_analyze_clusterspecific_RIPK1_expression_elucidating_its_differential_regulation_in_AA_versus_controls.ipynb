This notebook downloads the GEO dataset, pre-processes the data, and identifies cell clusters with distinct RIPK1 expression levels to understand its role in AA.

```python
import GEOparse
import scanpy as sc

gse = GEOparse.get_GEO('GSE269455')
adata = sc.read_10x_mtx('path_to_data')
sc.pp.filter_genes(adata, min_cells=3)
sc.tl.pca(adata, svd_solver='arpack')
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
sc.tl.umap(adata)
sc.pl.umap(adata, color=['RIPK1'], save='_RIPK1_expression.png')
```

This cell ranks differentially expressed genes across clusters to identify those with significant RIPK1 upregulation in AA.

```python
sc.tl.rank_genes_groups(adata, 'louvain', method='t-test')
sc.pl.rank_genes_groups(adata, n_genes=20, sharey=False, save='_ranked_genes.png')
```
