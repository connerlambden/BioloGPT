### Data Analysis of Flowering Intensity and Environmental Factors

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data for flowering intensity and environmental factors
data = {
    'Season': ['Summer', 'Monsoon', 'Winter'],
    'Flowering Intensity': [8.96, 8.48, 2.31],
    'Temperature': [30, 28, 15],
    'Humidity': [80, 75, 60]
}

df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 6))
sns.barplot(x='Season', y='Flowering Intensity', data=df)
plt.title('Flowering Intensity Across Seasons')
plt.xlabel('Season')
plt.ylabel('Flowering Intensity')
plt.show()
```
