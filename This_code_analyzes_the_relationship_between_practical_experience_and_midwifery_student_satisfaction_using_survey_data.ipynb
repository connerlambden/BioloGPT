### Step 1: Import necessary libraries and load data
import pandas as pd
import matplotlib.pyplot as plt

### Step 2: Load the survey data
# Assuming the data is in a CSV file
survey_data = pd.read_csv('midwifery_student_satisfaction.csv')

### Step 3: Analyze the data
# Calculate average satisfaction based on practical experience levels
average_satisfaction = survey_data.groupby('practical_experience_level')['satisfaction_score'].mean()

### Step 4: Visualize the results
plt.figure(figsize=(10, 6))
plt.bar(average_satisfaction.index, average_satisfaction.values)
plt.title('Average Satisfaction by Practical Experience Level')
plt.xlabel('Practical Experience Level')
plt.ylabel('Average Satisfaction Score')
plt.show()
