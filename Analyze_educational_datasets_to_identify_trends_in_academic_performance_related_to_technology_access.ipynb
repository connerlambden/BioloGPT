This notebook will analyze the impact of smartphone access on academic performance using relevant datasets.

```python
import pandas as pd

# Load dataset
dataset = pd.read_csv('education_data.csv')

# Analyze the impact of smartphone access on CGPA
results = dataset.groupby('smartphone_access')['CGPA'].mean()
```

The analysis will provide insights into how smartphone access correlates with academic performance.

```python
results.plot(kind='bar', title='Average CGPA by Smartphone Access')
```
