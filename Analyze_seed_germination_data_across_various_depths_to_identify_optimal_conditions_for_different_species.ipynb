This notebook will analyze the effects of planting depth on seed germination rates using collected data.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('seed_germination_data.csv')
plt.figure(figsize=(10,6))
for species in data['species'].unique():
    subset = data[data['species'] == species]
    plt.plot(subset['depth'], subset['germination_rate'], label=species)
plt.title('Seed Germination Rates by Planting Depth')
plt.xlabel('Planting Depth (cm)')
plt.ylabel('Germination Rate (%)')
plt.legend()
plt.show()
```

This analysis will help visualize the relationship between planting depth and germination rates for various species.
