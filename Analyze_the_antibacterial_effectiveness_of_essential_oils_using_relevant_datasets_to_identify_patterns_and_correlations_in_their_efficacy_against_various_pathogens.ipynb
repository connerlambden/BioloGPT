This notebook will analyze the antibacterial properties of essential oils using data extracted from various studies.

```python
import pandas as pd

dataset = pd.read_csv('essential_oils_data.csv')
# Analyze the dataset for antibacterial effectiveness.
```

The analysis will focus on identifying the most effective essential oils against specific bacterial strains.

```python
results = dataset.groupby('essential_oil')['MIC'].min()
results.to_csv('antibacterial_effectiveness.csv')
```

The results will provide insights into which essential oils are most effective against foodborne pathogens.

```python
print(results)
```
