This notebook demonstrates downloading a dataset, computing correlations between EAT-26 scores and future orientation measures, and visualizing the relationship.

```python
import pandas as pd
import matplotlib.pyplot as plt
url = 'https://osf.io/3y7dq/download'
data = pd.read_csv(url)
correlation = data['EAT26_score'].corr(data['future_orientation'])
print('Correlation between EAT26 and Future Orientation:', correlation)
plt.scatter(data['EAT26_score'], data['future_orientation'])
plt.title('EAT26 vs Future Orientation')
plt.xlabel('EAT26 Score')
plt.ylabel('Future Orientation Score')
plt.show()
```
