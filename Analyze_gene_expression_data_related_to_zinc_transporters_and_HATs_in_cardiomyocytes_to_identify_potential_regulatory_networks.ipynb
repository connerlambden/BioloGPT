This notebook will analyze gene expression data from cardiomyocytes under zinc overload conditions.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load gene expression data
expression_data = pd.read_csv('gene_expression_zinc.csv')

# Visualize expression levels of HATs and zinc transporters
sns.boxplot(data=expression_data, x='Condition', y='Expression', hue='Gene')
plt.title('Gene Expression Levels Under Zinc Overload')
plt.show()
```

The analysis will help identify which genes are significantly affected by zinc overload.

```python
# Statistical analysis to identify significant changes
from scipy import stats

# Perform t-test between control and zinc overload conditions
control = expression_data[expression_data['Condition'] == 'Control']['Expression']
zinc_overload = expression_data[expression_data['Condition'] == 'Zinc Overload']['Expression']
t_stat, p_value = stats.ttest_ind(control, zinc_overload)

print(f'T-statistic: {t_stat}, P-value: {p_value}')
```
