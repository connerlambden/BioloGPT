```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (replace with actual dataset URL or file path)
df = pd.read_csv('path_to_biopesticide_dataset.csv')

# Perform basic statistical analysis
summary = df.describe()
print(summary)

# Create a scatter plot for bioactivity vs concentration
plt.figure(figsize=(10,6))
plt.scatter(df['Concentration'], df['Bioactivity'], color='#6A0C76')
plt.xlabel('Concentration')
plt.ylabel('Bioactivity Index')
plt.title('Bioactivity vs Concentration of Isolated Compounds')
plt.show()
```

The code above performs statistical analysis and visualization to explore the relationship between compound concentration and bioactivity.
