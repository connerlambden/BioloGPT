This notebook will explore the gas sensing properties of WS2 and g-C3N4, focusing on their performance metrics and environmental stability.

```python
import pandas as pd

# Load relevant datasets
datasets = pd.read_csv('gas_sensing_data.csv')

# Analyze performance metrics
performance_metrics = datasets[['Material', 'Detection_Limit', 'Sensitivity']].groupby('Material').mean()
```

The analysis reveals the average detection limits and sensitivities for each material, guiding future sensor design.

```python
performance_metrics.to_csv('average_performance_metrics.csv')
```
