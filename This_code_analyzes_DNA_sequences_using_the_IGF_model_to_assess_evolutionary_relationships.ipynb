### Step 1: Import Required Libraries
Import necessary libraries for phylogenetic analysis.

```python
import numpy as np
import pandas as pd
from Bio import AlignIO
from Bio.Phylo import TreeConstruction
from Bio.Phylo import draw
```

### Step 2: Load DNA Sequence Data
Load your DNA sequence data in FASTA format.

```python
alignment = AlignIO.read('your_sequences.fasta', 'fasta')
```

### Step 3: Construct Phylogenetic Tree
Use the GTR+I+G+F model for tree construction.

```python
calculator = TreeConstruction.DistanceCalculator('identity')
dist_matrix = calculator.get_distance(alignment)
tree_constructor = TreeConstruction.NJTreeConstructor()
tree = tree_constructor.build_tree(dist_matrix)
```
