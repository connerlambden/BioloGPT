### Analyzing Antiviral Properties of Venom Toxins
This analysis will focus on the antiviral effects of various venom toxins against arboviruses.

```python
import pandas as pd

# Load relevant datasets
datasets = {
    'scorpine': {'EC50': 0.5, 'CC50': 5.0},
    'mucroporin_M1': {'EC50': 0.2, 'CC50': 3.0},
    'PLA2': {'EC50': 0.3, 'CC50': 4.0}
}

# Create a DataFrame
df = pd.DataFrame(datasets).T

# Calculate Selectivity Index (SI)
df['SI'] = df['CC50'] / df['EC50']

# Display the results
df
```
