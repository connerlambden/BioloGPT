This notebook loads the extracted IC50 and ACTH reduction data from the study, and generates grouped bar plots to visually compare the efficacy of Closantel and Niclosamide.

```python
import plotly.graph_objs as go
import plotly.offline as pyo

data = [
    {'compound': 'Closantel', 'IC50_USP8': 2.31, 'IC50_USP2': 16.8, 'ACTH_reduction': 60},
    {'compound': 'Niclosamide', 'IC50_USP8': None, 'IC50_USP2': None, 'ACTH_reduction': None}
]

compounds = [d['compound'] for d in data]
ic50 = [d['IC50_USP8'] for d in data]
acth = [d['ACTH_reduction'] for d in data]

trace1 = go.Bar(x=compounds, y=nic50, name='IC50 USP8 (µM)', marker=dict(color='#6A0C76'))
trace2 = go.Bar(x=compounds, y=acth, name='ACTH Reduction (%)', marker=dict(color='#FF5733'))

layout = go.Layout(title='USP8 Inhibition and ACTH Reduction Comparison', barmode='group')
fig = go.Figure(data=[trace1, trace2], layout=layout)
pyo.plot(fig, filename='usp8_inhibition.html')
```
