```python
import pandas as pd
import plotly.graph_objects as go

data = pd.read_csv('https://figshare.com/s/4f4b7bb5b38c117db768')
fig = go.Figure()
fig.add_trace(go.Scatter(x=data['time'], y=data['fluorescence'], mode='lines+markers', name='RNA Fluorescence'))
fig.update_layout(title='RNA Switch-RTF Dynamics', xaxis_title='Time (s)', yaxis_title='Fluorescence Intensity')
fig.show()
```
