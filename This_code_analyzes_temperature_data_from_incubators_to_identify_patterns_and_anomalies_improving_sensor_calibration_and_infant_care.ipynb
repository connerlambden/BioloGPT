### Data Analysis of Temperature Sensors in Incubators
This notebook will analyze temperature data collected from incubators to identify patterns and anomalies that can inform better sensor calibration.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('temperature_data.csv')
plt.figure(figsize=(10,5))
plt.plot(data['Time'], data['Temperature'], label='Temperature Over Time')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Data from Incubators')
plt.legend()
plt.show()
```
