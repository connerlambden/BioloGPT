### Data Analysis of Allelopathic Effects on Phytoplankton-Bacteria Interactions

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load experimental data
data = pd.read_csv('allelopathy_experiment_data.csv')

# Analyze growth rates of phytoplankton and bacteria
plt.figure(figsize=(10, 6))
plt.scatter(data['allelopathic_concentration'], data['phytoplankton_growth'], label='Phytoplankton Growth')
plt.scatter(data['allelopathic_concentration'], data['bacteria_growth'], label='Bacteria Growth')
plt.title('Impact of Allelopathic Compounds on Growth Rates')
plt.xlabel('Allelopathic Compound Concentration')
plt.ylabel('Growth Rate')
plt.legend()
plt.show()
```

### Discussion
This analysis will help elucidate the relationship between allelopathic compound concentration and the growth rates of phytoplankton and bacteria.
