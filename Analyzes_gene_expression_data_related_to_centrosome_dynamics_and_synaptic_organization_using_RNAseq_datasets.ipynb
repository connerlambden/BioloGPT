This notebook analyzes RNA-seq data to identify genes associated with centrosome dynamics and their impact on synaptic organization.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('centrosome_synaptic_data.csv')
sns.boxplot(x='Condition', y='Gene_Expression', data=data)
plt.title('Gene Expression in Different Conditions')
plt.show()
```
