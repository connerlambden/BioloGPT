This notebook analyzes data from social prescribing interventions to assess the impact of extended link worker support.

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('social_prescribing_data.csv')

# Analyze outcomes based on support duration
results = data.groupby('support_duration').agg({'wellbeing_score': 'mean', 'financial_outcome': 'sum'})
results
```

The results show the average wellbeing scores and total financial outcomes based on the duration of support.

```python
results.to_csv('analyzed_results.csv')
```
