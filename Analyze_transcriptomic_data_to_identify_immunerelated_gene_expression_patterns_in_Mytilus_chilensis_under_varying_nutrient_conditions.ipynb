This analysis will utilize RNA-Seq data from Mytilus chilensis to explore how nutrient levels influence immune gene expression.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load transcriptomic data
transcriptomic_data = pd.read_csv('path_to_data.csv')

# Filter for immune-related genes
immune_genes = transcriptomic_data[transcriptomic_data['gene_type'] == 'immune']

# Visualize expression levels
plt.figure(figsize=(10, 6))
sns.boxplot(x='nutrient_level', y='expression', data=immune_genes)
plt.title('Immune Gene Expression under Varying Nutrient Levels')
plt.show()
```

This code will help visualize the relationship between nutrient levels and immune gene expression in Mytilus chilensis.

```python
# Further analysis can include statistical tests to assess significance
from scipy import stats

# Example: ANOVA test for expression differences
anova_results = stats.f_oneway(immune_genes[immune_genes['nutrient_level'] == 'low']['expression'],
                                immune_genes[immune_genes['nutrient_level'] == 'medium']['expression'],
                                immune_genes[immune_genes['nutrient_level'] == 'high']['expression'])
print('ANOVA results:', anova_results)
```
