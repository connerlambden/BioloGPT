# Gene Expression Analysis of Animal vs Plant Cells
This notebook compares metabolic pathways in animal and plant cells using gene expression data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load gene expression data
animal_data = pd.read_csv('animal_cells_expression.csv')
plant_data = pd.read_csv('plant_cells_expression.csv')

# Analyze and visualize differences
sns.boxplot(data=[animal_data['metabolic_pathway'], plant_data['metabolic_pathway']])
plt.title('Comparison of Metabolic Pathways')
plt.show()
```

This analysis reveals key differences in metabolic pathways between animal and plant cells.
