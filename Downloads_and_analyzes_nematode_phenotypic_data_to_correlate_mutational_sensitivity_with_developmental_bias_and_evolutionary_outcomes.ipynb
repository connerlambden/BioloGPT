The notebook loads nematode phenotypic data from a public repository and utilizes statistical models to assess the relationship between mutation effects and cell fate sensitivities.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('https://github.com/Picao-Osorio/Evolution-Developmental-Bias/raw/main/DataS1.csv')

# Group by cell type and calculate mean mutation effect
summary = data.groupby('cell_type')['mutation_effect'].mean()

plt.figure(figsize=(8,5))
plt.bar(summary.index, summary.values, color='skyblue')
plt.xlabel('Cell Type')
plt.ylabel('Mean Mutation Effect')
plt.title('Mutational Sensitivity across Nematode Cell Types')
plt.show()
```

The above code segment visualizes mean mutation effects per cell type to highlight differences in mutational sensitivity.

```python
import statsmodels.api as sm
# Example: Fit a basic linear model (details would require further dataset context)
model_data = data[['mutation_effect', 'cell_type_index']].dropna()
X = sm.add_constant(model_data['cell_type_index'])
model = sm.OLS(model_data['mutation_effect'], X).fit()
print(model.summary())
```
