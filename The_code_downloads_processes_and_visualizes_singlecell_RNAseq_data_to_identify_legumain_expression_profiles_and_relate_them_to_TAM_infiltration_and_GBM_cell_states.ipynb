This notebook demonstrates the processing of scRNA-seq data from GBM samples, highlighting clusters with elevated legumain (LGMN) expression and correlating these with macrophage markers.

```python
import scanpy as sc
adata = sc.read_h5ad('GSE182109_processed.h5ad')
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)
adata.obs['LGMN_expression'] = adata[:, 'LGMN'].X
sc.pl.umap(adata, color='LGMN_expression', save='_LGMN_expression.png')
print('LGMN expression visualization complete.')
```

The above code filters the data, computes legumain expression, and generates a UMAP plot to visualize the spatial distribution of LGMN-positive cells.

```python
# Further steps might include differential expression analysis between high and low LGMN clusters
sc.tl.rank_genes_groups(adata, 'LGMN_high_low', method='wilcoxon')
sc.pl.rank_genes_groups(adata, n_genes=20, sharey=False)
print('Differential gene expression analysis complete.')
```
