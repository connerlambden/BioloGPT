This notebook will analyze RNA-seq data from FLASH-RT studies to identify differentially expressed genes.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load the dataset
data = pd.read_csv('flash_rt_data.csv')

# Perform differential expression analysis
results = stats.ttest_ind(data[data['treatment'] == 'FLASH']['expression'], data[data['treatment'] == 'CONV']['expression'])

# Output results
results_df = pd.DataFrame({'gene': data['gene'], 'p_value': results.pvalue})
results_df.to_csv('differential_expression_results.csv', index=False)
```
