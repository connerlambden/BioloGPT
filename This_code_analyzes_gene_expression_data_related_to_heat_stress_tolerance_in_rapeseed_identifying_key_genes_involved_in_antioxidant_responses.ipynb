This notebook analyzes gene expression data from rapeseed under heat stress conditions.

```python
import pandas as pd
# Load gene expression data
data = pd.read_csv('rapeseed_heat_stress.csv')
# Analyze key genes related to antioxidant responses.
```

The analysis focuses on identifying genes with significant changes in expression under heat stress.

```python
# Identify significant genes
significant_genes = data[data['p_value'] < 0.05]
```

This section summarizes the findings and discusses the implications for breeding heat-tolerant rapeseed.

```python
# Summary of findings
summary = significant_genes.groupby('gene_name').mean()
```
