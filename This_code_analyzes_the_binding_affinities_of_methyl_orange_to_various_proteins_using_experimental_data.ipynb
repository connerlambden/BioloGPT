### Binding Affinity Analysis of Methyl Orange to Proteins

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for binding affinities
data = {'Protein': ['Myosin', 'β-Lactoglobulin', 'Egg Albumin'], 'Binding Index': [2.4, 4.8, 2.5]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Protein'], df['Binding Index'], color=['blue', 'orange', 'green'])
plt.title('Binding Affinity of Methyl Orange to Proteins')
plt.xlabel('Protein Type')
plt.ylabel('Binding Index')
plt.show()
```

This analysis provides a visual representation of the binding affinities of methyl orange to different proteins.
