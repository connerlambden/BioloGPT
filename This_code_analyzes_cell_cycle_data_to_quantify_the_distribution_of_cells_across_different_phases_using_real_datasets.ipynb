This notebook will analyze cell cycle data from synchronized HeLa cells to quantify the distribution of cells in interphase and mitotic phases.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
cell_data = pd.read_csv('cell_cycle_data.csv')

# Analyze distribution
phase_counts = cell_data['phase'].value_counts()

# Plot results
phase_counts.plot(kind='bar')
plt.title('Cell Distribution Across Cell Cycle Phases')
plt.xlabel('Cell Cycle Phase')
plt.ylabel('Number of Cells')
plt.show()
```

This analysis provides insights into the distribution of cells across different phases of the cell cycle, which is crucial for understanding cell division dynamics.
