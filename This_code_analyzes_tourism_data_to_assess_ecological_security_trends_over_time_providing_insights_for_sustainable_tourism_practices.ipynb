### Analyzing Tourism Ecological Security Data
This notebook will analyze the ecological security data of tourism in Lanzhou from 2009 to 2021.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Year': [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'D_plus': [0.8, 0.809, 0.732, 0.68, 0.557, 0.513, 0.492, 0.585, 0.557, 0.526, 0.539, 0.567, 0.59],
    'D_minus': [0.402, 0.414, 0.483, 0.453, 0.553, 0.615, 0.617, 0.565, 0.569, 0.633, 0.7, 0.657, 0.598],
    'C': [0.334, 0.338, 0.398, 0.4, 0.499, 0.545, 0.556, 0.491, 0.506, 0.546, 0.565, 0.537, 0.503]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['D_plus'], marker='o', label='D+')
plt.plot(df['Year'], df['D_minus'], marker='o', label='D-')
plt.plot(df['Year'], df['C'], marker='o', label='C')
plt.title('Tourism Ecological Security Over Time')
plt.xlabel('Year')
plt.ylabel('Security Levels')
plt.legend()
plt.grid()
plt.show()
```

This analysis provides insights into the trends of ecological security in tourism, highlighting the need for sustainable practices.
