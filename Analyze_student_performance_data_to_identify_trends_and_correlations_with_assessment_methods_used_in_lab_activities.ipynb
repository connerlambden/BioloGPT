This notebook will analyze student performance data from various assessment methods used in lab activities.

```python
import pandas as pd

# Load student performance data
data = pd.read_csv('student_performance.csv')

# Analyze trends in performance based on assessment methods
performance_trends = data.groupby('assessment_method').mean()
performance_trends
```

The analysis provides insights into which assessment methods yield the best student performance.

```python
# Visualize the performance trends
import matplotlib.pyplot as plt

performance_trends.plot(kind='bar')
plt.title('Average Student Performance by Assessment Method')
plt.xlabel('Assessment Method')
plt.ylabel('Average Score')
plt.show()
```
