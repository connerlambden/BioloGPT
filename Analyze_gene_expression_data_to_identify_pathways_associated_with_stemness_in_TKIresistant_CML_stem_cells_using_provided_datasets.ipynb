## Analysis of Stemness Pathways in TKI-Resistant CML Stem Cells
This notebook analyzes gene expression data to identify key pathways associated with stemness in TKI-resistant CML stem cells.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

# Load gene expression data
expression_data = pd.read_csv('path_to_expression_data.csv')

# Identify stemness-related genes
stem_genes = ['SOX2', 'OCT3/4']

# Plot expression levels
plt.figure(figsize=(10,6))
for gene in stem_genes:
    sns.boxplot(x='Treatment', y=gene, data=expression_data, label=gene)
plt.title('Expression Levels of Stemness Genes after Imatinib Treatment')
plt.legend()
plt.show()
```

### Statistical Analysis
Performing t-tests to assess the significance of gene expression changes.

```python
for gene in stem_genes:
    treated = expression_data[expression_data['Treatment'] == 'Imatinib'][gene]
    control = expression_data[expression_data['Treatment'] == 'Control'][gene]
    stat, p = ttest_ind(treated, control)
    print(f'T-test for {gene}: p-value = {p}')
```
