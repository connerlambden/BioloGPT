Análisis de datos de savia para identificar correlaciones entre nutrientes y condiciones ambientales.

```python
import pandas as pd
# Cargar datos de savia
savia_data = pd.read_csv('savia_composicion.csv')
# Análisis de correlación
correlation = savia_data.corr()
```

Visualización de la correlación entre diferentes nutrientes en la savia.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(correlation, annot=True)
plt.title('Correlación entre nutrientes en la savia')
plt.show()
```
