This notebook downloads and compares neutron crystallography extracted data for the ligand-free, glucose complex, and 2F-Glc complex states of β-glucosidase.

```python
import plotly.graph_objects as go
import pandas as pd

data = {'Condition': ['Ligand-free','Glucose complex','2F-Glc complex'],
         'H_atoms': [3227, 3278, 3269],
         'D_atoms': [679, 815, 865],
         'Water_molecules': [346, 275, 304]}
df = pd.DataFrame(data)

fig = go.Figure(data=[
    go.Bar(name='H_atoms', x=df['Condition'], y=df['H_atoms']),
    go.Bar(name='D_atoms', x=df['Condition'], y=df['D_atoms']),
    go.Bar(name='Water_molecules', x=df['Condition'], y=df['Water_molecules'])
])

fig.update_layout(barmode='group', title='Neutron Crystallography Extracted Data for β-glucosidase States', plot_bgcolor='#F8F9FA')
fig.show()
```

The graph above visualizes the differences in hydrogen atoms, deuterium atoms, and water molecules across different enzyme states.
