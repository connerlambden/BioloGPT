This notebook downloads maize and rice QTL datasets and performs group LASSO regression to partition genetic contributions to heterosis.

```python
import pandas as pd
from sklearn.linear_model import Lasso
# Load dataset (replace with actual dataset paths)
maize_data = pd.read_csv('maize_dataset.csv')
rice_data = pd.read_csv('rice_dataset.csv')
# Perform LASSO regression to evaluate dominance effect
lasso = Lasso(alpha=0.1)
X = maize_data.drop('phenotype', axis=1)
y = maize_data['phenotype']
lasso.fit(X, y)
print(lasso.coef_)
# Similar analysis for rice data follows.
```

The above code outlines the preliminary steps to analyze the contributions of individual QTL to the observed heterosis, demonstrating the utility of regularized regression models.
