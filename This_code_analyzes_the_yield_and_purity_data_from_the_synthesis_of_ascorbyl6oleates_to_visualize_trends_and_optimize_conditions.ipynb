### Analysis of Synthesis Yields and Purity of Ascorbyl-6-Oleates
This notebook will analyze the data from the synthesis of ascorbyl-6-oleates, focusing on the yields and purity achieved under different conditions.

```python
import pandas as pd
import plotly.express as px

data = {
    'Method': ['Esterification', 'Transesterification'],
    'Yield (%)': [19.7, 73.8],
    'Purity (%)': [94.8, 94.3]
}
df = pd.DataFrame(data)

fig = px.bar(df, x='Method', y='Yield (%)', title='Synthesis Yields of Ascorbyl-6-Oleates', color='Yield (%)', text='Yield (%)')
fig.show()
```
