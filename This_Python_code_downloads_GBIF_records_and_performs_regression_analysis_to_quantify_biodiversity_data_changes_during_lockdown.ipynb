Load and pre-process GBIF and mobility datasets for regression analysis.

```python
import pandas as pd
import numpy as np
from nlme import gls  # hypothetical import

# Load data
df_gbif = pd.read_csv('path_to_gbif_data.csv')
df_mobility = pd.read_csv('path_to_mobility_data.csv')

# Data transformation
df_gbif['log_records'] = np.log10(df_gbif['records'] + 100)

# Merge datasets and perform regression analysis
merged_df = pd.merge(df_gbif, df_mobility, on='country')
model = gls('log_records ~ stringency + mobility_change', data=merged_df)
results = model.fit()
print(results.summary())
```
