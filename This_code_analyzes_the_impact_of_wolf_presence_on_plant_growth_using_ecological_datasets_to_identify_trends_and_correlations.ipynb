This notebook will analyze the impact of wolf presence on plant growth using ecological datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
aspen_data = pd.read_csv('aspen_growth_data.csv')

# Analyze trends
plt.plot(aspen_data['Year'], aspen_data['Height'], label='Aspen Height')
plt.title('Aspen Growth Over Time')
plt.xlabel('Year')
plt.ylabel('Height (cm)')
plt.legend()
plt.show()
```

The analysis will visualize the trends in aspen growth over the years, correlating with wolf presence.
