### Force-Velocity Analysis of Molecular Motors

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for molecular motors
motor_data = {'Motor': ['Kinesin-1', 'RNA Polymerase', 'Bacterial Flagellar Motor'],
              'Stall Force (pN)': [5, 25, 150],
              'Velocity (nm/s)': [800, 200, 1000]}
df = pd.DataFrame(motor_data)

# Plotting the data
plt.bar(df['Motor'], df['Stall Force (pN)'], color='blue')
plt.title('Stall Forces of Molecular Motors')
plt.xlabel('Molecular Motor')
plt.ylabel('Stall Force (pN)')
plt.show()
```
