### Survey Data Analysis on Moon Cake Delivery Visibility

```python
import pandas as pd

# Load survey data
survey_data = pd.read_csv('moon_cake_delivery_survey.csv')

# Analyze the effectiveness of different light sources
light_effectiveness = survey_data.groupby('light_source')['visibility_rating'].mean()
print(light_effectiveness)
```
