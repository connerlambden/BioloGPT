This notebook section outlines the import and analysis of experimental calcium imaging data from electrically stunned zebrafish larvae. The goal is to quantify change in calcium signal amplitude over time.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Code to download and load dataset
url = 'https://biostudies.example.org/dataset/zebrafish_calcium.csv'
data = pd.read_csv(url)
print(data.head())
```

Next, plot the calcium signal dynamics for different time points post-stunning to visually inspect the propagation and decay of neural activity.

```python
plt.figure(figsize=(10, 6))
plt.plot(data['time'], data['calcium_signal'], marker='o')
plt.title('Calcium Signal Dynamics Post Electrical Stunning')
plt.xlabel('Time (seconds)')
plt.ylabel('Calcium Signal Amplitude')
plt.grid(True)
plt.show()
```

This simple analysis can be extended with statistical testing to compare pre- and post-stunning signal distributions.

```python
from scipy.stats import ttest_ind
pre_stun = data[data['time'] < 1]['calcium_signal']
post_stun = data[data['time'] >= 1]['calcium_signal']
stat, p = ttest_ind(pre_stun, post_stun)
print(f'T-test statistic: {stat}, p-value: {p}')
```
