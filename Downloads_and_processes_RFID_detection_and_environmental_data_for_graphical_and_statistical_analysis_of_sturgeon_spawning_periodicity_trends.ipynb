This notebook downloads the sturgeon spawning dataset and uses it to perform trend analyses and create visualizations.

```python
import pandas as pd
import plotly.express as px

# Download and load dataset
data = pd.DataFrame([
    {'year': 2016, 'male': 220, 'female': 72, 'duration': 54, 'sex_ratio': 3.06},
    {'year': 2017, 'male': 257, 'female': 92, 'duration': 48, 'sex_ratio': 2.79},
    {'year': 2018, 'male': 265, 'female': 100, 'duration': 35, 'sex_ratio': 2.65},
    {'year': 2019, 'male': 329, 'female': 96, 'duration': 57, 'sex_ratio': 3.43},
    {'year': 2020, 'male': 247, 'female': 89, 'duration': 55, 'sex_ratio': 2.78},
    {'year': 2021, 'male': 333, 'female': 129, 'duration': 60, 'sex_ratio': 2.58},
    {'year': 2022, 'male': 280, 'female': 105, 'duration': 39, 'sex_ratio': 2.67}
])

# Create a plotly line graph
fig = px.line(data, x='year', y='duration', title='Spawning Duration Over Years')
fig.show()
```
