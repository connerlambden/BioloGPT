Load TMB, OS, and PFS data from the study and visualize hazard ratios via a grouped bar chart.

```python
import pandas as pd
import plotly.express as px

data = pd.DataFrame({'Study': ['Study 1', 'Study 2'],
                      'OS_HR': [0.56, 0.36],
                      'PFS_HR': [0.5, 0.44]})

fig = px.bar(data, x='Study', y=['OS_HR', 'PFS_HR'], barmode='group', 
             title='TMB Impact on OS and PFS', 
             labels={'value':'Hazard Ratio (HR)', 'variable':'Survival Metric'})
fig.show()
```
