This notebook will analyze the impact of probiotic supplementation on gut microbiota composition using 16S rRNA sequencing data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('probiotic_study_data.csv')

# Analyze microbiota composition
composition = data.groupby('group').mean()

# Visualize results
composition.plot(kind='bar')
plt.title('Gut Microbiota Composition Changes')
plt.ylabel('Relative Abundance')
plt.show()
```

The analysis provides insights into how different probiotic strains affect gut microbiota composition.
