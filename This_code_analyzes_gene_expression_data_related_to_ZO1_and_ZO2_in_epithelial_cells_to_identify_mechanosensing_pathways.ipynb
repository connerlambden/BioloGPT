Analyze gene expression data for ZO-1 and ZO-2.

```python
import pandas as pd
# Load gene expression data
data = pd.read_csv('gene_expression_data.csv')
# Filter for ZO-1 and ZO-2
data_filtered = data[data['gene'].isin(['ZO-1', 'ZO-2'])]
```

Perform statistical analysis to identify significant changes in expression.

```python
from scipy import stats
# Perform t-test
stat, p_value = stats.ttest_ind(data_filtered['expression'][data_filtered['condition'] == 'control'], data_filtered['expression'][data_filtered['condition'] == 'treatment'])
print(f'Statistic: {stat}, p-value: {p_value}')
```
