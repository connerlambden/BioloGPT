This notebook calculates the relative molecular mass of air using the ideal gas law.

```python
# Constants
mass = 0.3  # in grams
volume = 0.240  # in liters
R = 0.0821  # L·atm/(K·mol)
T = 273.15  # in Kelvin
P = 1  # in atm

# Calculate molar mass
molar_mass = (mass * R * T) / (P * volume)
molar_mass
```

The calculated molar mass of air is approximately 28.97 g/mol.

```python
# Display the result
print(f'Molar Mass of Air: {molar_mass:.2f} g/mol')
```
