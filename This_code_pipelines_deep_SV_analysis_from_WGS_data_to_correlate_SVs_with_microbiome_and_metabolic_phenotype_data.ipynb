Step 1: Load the WGS-derived SV dataset and the associated microbiome/metabolite phenotype tables for the 4D-SZ cohort.

```python
import pandas as pd
# Load SV dataset
sv_data = pd.read_csv('4D_SZ_SV_data.csv')
# Load microbial features and metabolic traits
micro_data = pd.read_csv('4D_SZ_microbiome.csv')
metabolite_data = pd.read_csv('4D_SZ_metabolites.csv')

print(sv_data.head(), micro_data.head(), metabolite_data.head())
```

Step 2: Perform association analysis between common SVs and microbial features/metabolic traits using linear regression.

```python
import statsmodels.api as sm

# Example: Association analysis for a specific SV
sv_feature = sv_data['SV_123']  # hypothetical SV indicator
micro_feature = micro_data['Coprobacillus_sp_29_1']

# Add intercept term
X = sm.add_constant(sv_feature)
model = sm.OLS(micro_feature, X).fit()
print(model.summary())
```

Step 3: Conduct Mendelian randomization using identified SVs as instruments to assess causality on metabolic traits.

```python
# This is a sketch; in practice use specialized MR packages 
import numpy as np

def perform_mr(instrument, exposure, outcome):
    beta_iv = np.corrcoef(instrument, outcome)[0,1] / np.corrcoef(instrument, exposure)[0,1]
    return beta_iv

beta_iv = perform_mr(sv_data['SV_123'], micro_data['Coprobacillus_sp_29_1'], metabolite_data['MCV'])
print('Estimated causal effect:', beta_iv)
```

Step 4: Visualize the significant associations using Plotly.

```python
import plotly.express as px

fig = px.scatter(x=sv_data['SV_123'], y=metabolite_data['MCV'], 
                 title='SV 123 vs. MCV Association', labels={'x':'SV 123','y':'Mean Corpuscular Volume'})
fig.show()
```
