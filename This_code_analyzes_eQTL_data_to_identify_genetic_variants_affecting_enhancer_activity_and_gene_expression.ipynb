### Step 1: Import Libraries
Import necessary libraries for data analysis and visualization.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2: Load eQTL Data
Load the eQTL dataset containing genetic variants and their associations with enhancer activity.

```python
eqtl_data = pd.read_csv('eqtl_data.csv')
print(eqtl_data.head())
```

### Step 3: Analyze Associations
Analyze the associations between genetic variants and enhancer activity.

```python
association_results = eqtl_data.groupby('variant').agg({'enhancer_activity':'mean'})
print(association_results)
```
