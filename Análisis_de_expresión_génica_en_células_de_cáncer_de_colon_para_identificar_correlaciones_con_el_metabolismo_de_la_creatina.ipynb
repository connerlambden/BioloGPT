Este análisis utiliza datos de expresión génica para investigar la relación entre la creatina y el cáncer de colon.

```python
import pandas as pd
# Cargar datos de expresión génica
expression_data = pd.read_csv('data/colon_cancer_expression.csv')
# Análisis de correlación
correlation = expression_data.corr()
```

Los resultados del análisis de correlación ayudarán a identificar genes que están relacionados con el metabolismo de la creatina en el contexto del cáncer de colon.

```python
# Visualización de la matriz de correlación
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(correlation, annot=True)
plt.title('Matriz de Correlación de Expresión Génica')
plt.show()
```
