### Data Analysis of Alzheimer's Disease Biomarkers
This notebook analyzes the variability of plasma biomarkers in Alzheimer's disease patients.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for biomarkers
biomarkers_data = {
    'Biomarker': ['Aβ40', 'Aβ42', 'p-tau181', 'NfL', 'GFAP'],
    'Mean Concentration (pg/mL)': [103.2, 5.4, 8.8, 31.4, 128.9],
    'Intra-individual Variability (%)': [5.2, 5.9, 7.0, 7.3, 11.2],
    'Inter-individual Variability (%)': [10.0, 14.7, 20.5, 34.1, 20.5]
}

# Create DataFrame
biomarkers_df = pd.DataFrame(biomarkers_data)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(biomarkers_df['Biomarker'], biomarkers_df['Inter-individual Variability (%)'], color='orange', label='Inter-individual Variability')
plt.bar(biomarkers_df['Biomarker'], biomarkers_df['Intra-individual Variability (%)'], color='blue', alpha=0.7, label='Intra-individual Variability')
plt.title('Variability of Alzheimer’s Disease Biomarkers')
plt.xlabel('Biomarkers')
plt.ylabel('Variability (%)')
plt.legend()
plt.show()
```

This analysis provides insights into the variability of biomarkers, which is crucial for their clinical application.
