### Data Analysis of Biomarkers in SAH Patients

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from the study
data = {
    'Group': ['Control', 'CVS Day 1', 'CVS Day 4', 'CVS Day 7', 'CVS Day 10'],
    'IL-6 (pg/ml)': [4.01, 10.65, 29.26, 37.93, 52.39],
    'CRP (mg/l)': [1.15, 5.99, 13.83, 15.17, 25.82],
    'S100 (µg/l)': [0.25, 0.53, 0.81, 1.03, 1.71],
    'NO (µmol/l)': [2.31, 4.14, 5.97, 7.82, 8.85]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
for column in df.columns[1:]:
    plt.plot(df['Group'], df[column], marker='o', label=column)
plt.title('Biomarker Levels in SAH Patients')
plt.xlabel('Patient Group')
plt.ylabel('Levels')
plt.legend()
plt.grid()
plt.show()
```
