### Data Description
This section downloads and processes SMN copy number data, visualizing the relationship between SMN1 and SMN2 to reflect the gene conversion events observed in the study.

```python
import pandas as pd
import plotly.express as px

# Define the dataset based on extracted copy number data
data = [
    {'SMN1_copy_number': 4, 'SMN2_copy_number': 0, 'count': 1},
    {'SMN1_copy_number': 4, 'SMN2_copy_number': 1, 'count': 1},
    {'SMN1_copy_number': 3, 'SMN2_copy_number': 1, 'count': 9},
    {'SMN1_copy_number': 3, 'SMN2_copy_number': 2, 'count': 2},
    {'SMN1_copy_number': 2, 'SMN2_copy_number': 0, 'count': 10},
    {'SMN1_copy_number': 2, 'SMN2_copy_number': 1, 'count': 61},
    {'SMN1_copy_number': 2, 'SMN2_copy_number': 2, 'count': 90},
    {'SMN1_copy_number': 2, 'SMN2_copy_number': 3, 'count': 3},
    {'SMN1_copy_number': 1, 'SMN2_copy_number': 0, 'count': 1},
    {'SMN1_copy_number': 1, 'SMN2_copy_number': 1, 'count': 1},
    {'SMN1_copy_number': 1, 'SMN2_copy_number': 2, 'count': 1},
    {'SMN1_copy_number': 0, 'SMN2_copy_number': 1, 'count': 3},
    {'SMN1_copy_number': 0, 'SMN2_copy_number': 2, 'count': 13},
    {'SMN1_copy_number': 0, 'SMN2_copy_number': 3, 'count': 1},
    {'SMN1_copy_number': 0, 'SMN2_copy_number': 4, 'count': 1}
]

df = pd.DataFrame(data)

# Create scatter plot to visualize the inverse correlation
fig = px.scatter(df, x='SMN1_copy_number', y='SMN2_copy_number', size='count',
                 title='Inverse Correlation Between SMN1 and SMN2 Copy Numbers')
fig.show()
```

### Statistical Analysis
The following code performs a Pearson correlation analysis to statistically validate the inverse correlation between SMN1 and SMN2.

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(df['SMN1_copy_number'], df['SMN2_copy_number'])
print(f"Pearson correlation: {corr}, p-value: {p_value}")
```
