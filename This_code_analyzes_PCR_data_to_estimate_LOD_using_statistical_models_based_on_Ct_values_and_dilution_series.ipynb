### PCR LOD Analysis
This notebook analyzes PCR data to estimate the Limit of Detection (LOD) using statistical models.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load PCR data
pcr_data = pd.read_csv('pcr_data.csv')

# Calculate LOD based on Ct values
lod_estimate = np.percentile(pcr_data['Ct'], 95)  # 95% detection threshold
print(f'Estimated LOD: {lod_estimate}')
```

### Results
The estimated LOD is calculated based on the 95% detection threshold from the Ct values.

```python
# Plotting the results
plt.figure(figsize=(10, 6))
plt.hist(pcr_data['Ct'], bins=20, alpha=0.7, color='blue')
plt.axvline(lod_estimate, color='red', linestyle='dashed', linewidth=2)
plt.title('Distribution of Ct Values')
plt.xlabel('Ct Value')
plt.ylabel('Frequency')
plt.show()
```
