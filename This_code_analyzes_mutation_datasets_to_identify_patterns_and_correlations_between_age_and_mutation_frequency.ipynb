### Analyzing DNA Mutation Data
This notebook will explore the relationship between age and the frequency of DNA mutations in different tissues.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
age = [20, 25, 30, 35, 40]
oocyte_mutations = [5, 4, 6, 3, 2]
blood_mutations = [10, 12, 15, 20, 25]
saliva_mutations = [3, 2, 4, 5, 6]

# Create DataFrame
mutation_data = pd.DataFrame({'Age': age, 'Oocyte': oocyte_mutations, 'Blood': blood_mutations, 'Saliva': saliva_mutations})

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(mutation_data['Age'], mutation_data['Oocyte'], marker='o', label='Oocyte Mutations')
plt.plot(mutation_data['Age'], mutation_data['Blood'], marker='o', label='Blood Mutations')
plt.plot(mutation_data['Age'], mutation_data['Saliva'], marker='o', label='Saliva Mutations')
plt.title('Accumulation of DNA Mutations with Age')
plt.xlabel('Age (years)')
plt.ylabel('Number of Mutations')
plt.legend()
plt.grid()
plt.show()
```

This analysis provides insights into how mutation frequencies vary with age across different tissues.
