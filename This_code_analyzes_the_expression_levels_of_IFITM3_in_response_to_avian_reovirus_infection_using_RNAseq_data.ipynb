### Analyzing IFITM3 Expression in Avian Reovirus Infection

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load RNA-seq data
rna_seq_data = pd.read_csv('path_to_rna_seq_data.csv')

# Filter for IFITM3 expression levels
ifitm3_expression = rna_seq_data[rna_seq_data['gene'] == 'IFITM3']

# Plotting the expression levels
plt.figure(figsize=(10,6))
plt.bar(ifitm3_expression['time_point'], ifitm3_expression['expression_level'])
plt.title('IFITM3 Expression Levels During Avian Reovirus Infection')
plt.xlabel('Time Point')
plt.ylabel('Expression Level')
plt.show()
```

This analysis provides insights into how IFITM3 expression changes over time during avian reovirus infection, helping to elucidate its role in the antiviral response.
