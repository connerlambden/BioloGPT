This notebook block describes the data download steps from the provided clinical trial datasets, setup for survival analysis, and subsequent plotting of Kaplan-Meier estimates to compare trial outcomes using CURE AI-driven enrollment.

```python
import pandas as pd
import lifelines
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt

# Example: Load clinical trial data (real dataset link to be provided by source)
trial_data = pd.read_csv('clinical_trial_data.csv')

# Fit the Kaplan Meier model
kmf = KaplanMeierFitter()
fig, ax = plt.subplots(figsize=(8, 6))

# Assume 'time', 'event', and 'group' columns exist
for label, grouped_df in trial_data.groupby('group'):
    kmf.fit(grouped_df['time'], event_observed=grouped_df['event'], label=label)
    kmf.plot_survival_function(ax=ax)

plt.title('Kaplan-Meier Survival Estimates by CURE AI Enriched Groups')
plt.xlabel('Time (days)')
plt.ylabel('Survival Probability')
plt.show()
```

This markdown cell explains the code: it loads the clinical trial dataset, groups the patients by the predicted benefit based on the CURE Index, fits a Kaplan-Meier estimator for each group, and plots the survival curves.

```python
# Additional analysis: Log-rank test for group differences
from lifelines.statistics import logrank_test

# Compare two statistically enriched groups (example: group A vs group B)
group_A = trial_data[trial_data['group'] == 'A']
group_B = trial_data[trial_data['group'] == 'B']

results = logrank_test(group_A['time'], group_B['time'], event_observed_A=group_A['event'], event_observed_B=group_B['event'])
print('P-value from log-rank test:', results.p_value)
```

This final block performs a statistical test (log-rank test) to assess the differences in survival distributions between the enriched groups, providing quantitative evidence of treatment effect differences.
