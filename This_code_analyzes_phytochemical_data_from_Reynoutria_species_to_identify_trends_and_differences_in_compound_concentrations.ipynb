This notebook will analyze the phytochemical content of Reynoutria species, focusing on stilbenes and anthraquinones.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Species': ['R. japonica (China)', 'R. japonica (Poland)', 'R. sachalinensis'],
        'Piceid': [14.83, 7.45, 4.67],
        'Resveratrol': [1.29, 0.65, 0.52],
        'Emodin': [4.93, 4.01, 1.93]}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
for compound in df.columns[1:]:
    plt.bar(df['Species'], df[compound], label=compound)
plt.title('Phytochemical Content in Reynoutria Species')
plt.ylabel('Concentration (mg/g)')
plt.legend()
plt.show()
```

This analysis provides insights into the differences in phytochemical profiles among Reynoutria species.
