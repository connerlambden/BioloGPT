This notebook loads the dataset containing seroprevalence percentages and risk factor odds ratios, then creates plots to visualize the key metrics.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.DataFrame({
    'Agent': ['Neosporosis', 'Toxoplasmosis', 'Chlamydiosis', 'Q Fever', 'BVD', 'IBR'],
    'Seroprevalence': [31.57, 28.94, 15.78, 47.36, 60.52, 42.10]
})

# Plotting
plt.figure(figsize=(10,6))
plt.bar(data['Agent'], data['Seroprevalence'], color='#6A0C76')
plt.xlabel('Infectious Agents')
plt.ylabel('Seroprevalence (%)')
plt.title('Seroprevalence of Infectious Agents in Cattle')
plt.show()
```

This cell explains the workflow for integrating seroprevalence data with risk statistics to further explore associations via regression analysis.

```python
import statsmodels.api as sm

# Sample risk factors data
data_risk = pd.DataFrame({
    'Agent': ['Neosporosis', 'Toxoplasmosis', 'IBR', 'BVD', 'Q Fever'],
    'OR': [1.11, 1.95, 1.78, 1.65, 1.51]
})

# Dummy predictor variable
X = sm.add_constant(data_risk['OR'])
# Dummy outcome variable
Y = data_risk['OR'] * 0.5  # Simplified
model = sm.OLS(Y, X).fit()
print(model.summary())
```
