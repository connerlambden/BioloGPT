### Step 1: Data Preparation
Load the relevant blaTEM gene sequences from GenBank and prepare them for analysis.

```python
from Bio import SeqIO

# Load blaTEM sequences from GenBank
sequences = list(SeqIO.parse('blatem_sequences.gb', 'genbank'))
```

### Step 2: Variant Analysis
Identify and categorize variants of the blaTEM gene based on sequence differences.

```python
from Bio.Seq import Seq
from Bio.SeqUtils import GC

variants = {}
for seq_record in sequences:
    variant_id = seq_record.id
    sequence = str(seq_record.seq)
    gc_content = GC(sequence)
    variants[variant_id] = {'sequence': sequence, 'gc_content': gc_content}
```
