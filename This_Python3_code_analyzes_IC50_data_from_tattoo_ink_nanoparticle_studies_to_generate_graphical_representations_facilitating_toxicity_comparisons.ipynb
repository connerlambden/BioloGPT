This section downloads the relevant datasets and prepares the analysis environment.

```python
import pandas as pd
import plotly.express as px
# Load dataset with nanoparticle toxicity data
# DataFrame includes columns: NP_type, IC50, Cell_Type
df = pd.DataFrame([
    {'NP_type': 'CuO', 'IC50': 17, 'Cell_Type': 'HaCaT'},
    {'NP_type': 'CuO', 'IC50': 1.2, 'Cell_Type': 'Primary Fibroblasts'},
    {'NP_type': 'ZnO', 'IC50': 50, 'Cell_Type': 'Primary Fibroblasts'},
    {'NP_type': 'ZnO', 'IC50': 10, 'Cell_Type': 'Primary Keratinocytes'}
])
fig = px.bar(df, x='Cell_Type', y='IC50', color='NP_type', barmode='group', title='Nanoparticle IC50 Values')
fig.show()
```
