This notebook downloads experimental efficiency datasets from Zenodo and processes genome editing data from different Dictyostelid species.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('https://zenodo.org/record/15039721/files/editing_efficiencies.csv')
plt.figure(figsize=(10,6))
for species in data['species'].unique():
    subset = data[data['species'] == species]
    plt.plot(subset['experiment_day'], subset['efficiency'], label=species)
plt.xlabel('Days Post-Transfection')
plt.ylabel('Editing Efficiency (%)')
plt.title('CRISPR-Cas9 Editing Efficiencies Across Dictyostelid Species')
plt.legend()
plt.show()
```

The above plot visualizes the temporal dynamics of genome editing efficiency across various Dictyostelid species, providing insights into protocol robustness.

```python
import seaborn as sns

sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.barplot(x='species', y='efficiency', data=data, palette='deep')
plt.xlabel('Species')
plt.ylabel('Efficiency (%)')
plt.title('Comparison of Editing Efficiencies by Species')
plt.xticks(rotation=45)
plt.show()
```

These visualizations help in comparing the effectiveness of the new editing protocol across different species, highlighting its potential for broad application.
