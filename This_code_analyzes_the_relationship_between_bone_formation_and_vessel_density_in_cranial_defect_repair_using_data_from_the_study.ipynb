### Analysis of Bone Formation and Vessel Density
This notebook will analyze the data on new bone formation and vessel density from the study, providing insights into the relationship between these two factors during cranial defect repair.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from the study
data = {'Time Point': ['Week 3', 'Week 5'],
        'Superior Periosteum Bone Formation (%)': [11, 3.4],
        'Dura Periosteum Bone Formation (%)': [61, 3],
        'Dura Periosteum Vessel Density (fold increase)': [5, 5]}  

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['Time Point'], df['Superior Periosteum Bone Formation (%)'], label='Superior Periosteum')
plt.bar(df['Time Point'], df['Dura Periosteum Bone Formation (%)'], label='Dura Periosteum', bottom=df['Superior Periosteum Bone Formation (%)'])
plt.title('Bone Formation Comparison')
plt.ylabel('Bone Formation (%)')
plt.legend()
plt.show()
```
