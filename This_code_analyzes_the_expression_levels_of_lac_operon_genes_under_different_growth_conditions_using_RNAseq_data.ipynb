### RNA-seq Data Analysis for lac Operon Expression

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load RNA-seq data
data = pd.read_csv('lac_operon_expression.csv')

# Plot expression levels
plt.figure(figsize=(10,6))
plt.bar(data['Condition'], data['Expression_Level'])
plt.title('Lac Operon Gene Expression Under Different Conditions')
plt.xlabel('Condition')
plt.ylabel('Expression Level')
plt.show()
```

This analysis provides insights into how different conditions affect the expression of the lac operon genes.
