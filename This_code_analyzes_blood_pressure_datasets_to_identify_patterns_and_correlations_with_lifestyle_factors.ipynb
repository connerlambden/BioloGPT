This notebook will explore the relationship between blood pressure and various lifestyle factors using real-world datasets.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
bp_data = pd.read_csv('blood_pressure_data.csv')

# Analyze correlations
correlation_matrix = bp_data.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation between Lifestyle Factors and Blood Pressure')
plt.show()
```

The analysis reveals significant correlations between dietary habits, physical activity, and blood pressure levels.

```python
# Further analysis on specific groups
hypertensive_patients = bp_data[bp_data['hypertension'] == 1]
# Visualize data
sns.boxplot(x='age_group', y='blood_pressure', data=hypertensive_patients)
plt.title('Blood Pressure Distribution by Age Group')
plt.show()
```
