### Data Analysis of Caspase Expression Levels
This notebook will analyze the expression levels of caspase 3 and cleaved caspase 3 in various datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
healthy_data = pd.read_csv('healthy_cells.csv')
cancer_data = pd.read_csv('cancer_cells.csv')

# Analyze expression levels
healthy_mean = healthy_data.mean()
cancer_mean = cancer_data.mean()

# Plotting
plt.bar(['Healthy', 'Cancer'], [healthy_mean['caspase3'], cancer_mean['caspase3']], color=['blue', 'red'])
plt.title('Caspase 3 Expression Levels')
plt.ylabel('Expression Level')
plt.show()
```
