This notebook will analyze RNA-seq data to identify genes associated with nucleolar functions in cancer.

```python
import pandas as pd
import numpy as np

# Load RNA-seq data
data = pd.read_csv('rna_seq_data.csv')

# Analyze differentially expressed genes
results = data[data['p_value'] < 0.05]
```

The analysis identifies genes significantly associated with nucleolar functions, providing insights into their roles in cancer.

```python
results.to_csv('differentially_expressed_genes.csv', index=False)
```
