This notebook loads gene expression datasets corresponding to soft and stiff substrate conditions and performs t-tests to identify genes significantly altered, thereby linking substrate mechanics to receptor activity.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load gene expression datasets for soft and stiff substrates
data_soft = pd.read_csv('soft_substrate_data.csv')
data_stiff = pd.read_csv('stiff_substrate_data.csv')

# Perform differential expression analysis
results = {}
for gene in data_soft.columns[1:]:
    stat, pval = stats.ttest_ind(data_soft[gene], data_stiff[gene])
    results[gene] = pval

# Convert results to DataFrame and filter by significance
df_results = pd.DataFrame(list(results.items()), columns=['gene', 'p_value'])
df_significant = df_results[df_results['p_value'] < 0.05]
print(df_significant.head())
```

The output provides a list of candidate genes whose expression is significantly impacted by substrate stiffness, potentially including GPCR-related signaling mediators.
