### Simulating Fragment Generation and Size Analysis
This notebook will simulate the generation of DNA fragments and analyze their sizes based on specified parameters.

```python
import numpy as np
import pandas as pd

# Simulate fragment sizes
np.random.seed(42)
fragment_sizes = np.random.normal(loc=200, scale=50, size=1000)  # mean size 200bp, std dev 50bp

# Create a DataFrame to analyze sizes
fragment_df = pd.DataFrame(fragment_sizes, columns=['Fragment Size'])

# Summary statistics
summary_stats = fragment_df.describe()
summary_stats
```
