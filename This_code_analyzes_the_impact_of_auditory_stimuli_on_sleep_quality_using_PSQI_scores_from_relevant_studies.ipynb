### Auditory Stimuli and Sleep Quality Analysis
This notebook analyzes the impact of different auditory stimuli on sleep quality using PSQI scores.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for PSQI scores
data = {'Condition': ['Ambient Sound A', 'Ambient Sound B', 'Baseline'],
        'PSQI Score': [8.16, 7.5, 9.0]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Condition'], df['PSQI Score'], color=['blue', 'green', 'red'])
plt.title('Average Global PSQI Scores')
plt.xlabel('Sound Condition')
plt.ylabel('PSQI Score')
plt.show()
```
