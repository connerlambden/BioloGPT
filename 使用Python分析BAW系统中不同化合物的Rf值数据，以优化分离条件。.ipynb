分析BAW系统中不同化合物的Rf值数据，优化分离条件。

```python
import pandas as pd
# 读取Rf值数据
data = pd.read_csv('Rf_values.csv')
# 计算不同溶剂条件下的Rf值平均值
mean_rf = data.groupby('solvent')['Rf'].mean()
```
