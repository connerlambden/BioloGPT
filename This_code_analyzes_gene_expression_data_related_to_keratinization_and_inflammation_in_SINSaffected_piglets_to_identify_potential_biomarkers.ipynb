### Gene Expression Analysis for SINS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_sins.csv')
sns.boxplot(x='gene', y='expression', data=data)
plt.title('Gene Expression in SINS-affected Piglets')
plt.show()
```
