This notebook will analyze tweets and posts related to 'Red China' to determine public sentiment.

```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

def analyze_sentiment(data):
    nltk.download('vader_lexicon')
    sia = SentimentIntensityAnalyzer()
    data['sentiment'] = data['text'].apply(lambda x: sia.polarity_scores(x)['compound'])
    return data

# Load data
# data = pd.read_csv('social_media_data.csv')
# analyzed_data = analyze_sentiment(data)
```
