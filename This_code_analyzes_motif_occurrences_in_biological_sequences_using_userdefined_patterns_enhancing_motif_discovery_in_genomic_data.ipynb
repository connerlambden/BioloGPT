### Step 1: Import Required Libraries
Import necessary libraries for sequence analysis and motif searching.

```python
import pandas as pd
from Bio import SeqIO
import re

# Load sequences
sequences = SeqIO.to_dict(SeqIO.parse('sequences.fasta', 'fasta'))
```

### Step 2: Define Motif Search Function
Create a function to search for user-defined motifs in the sequences.

```python
def find_motifs(sequences, motif):
    results = {}
    for seq_id, seq in sequences.items():
        matches = re.findall(motif, str(seq.seq))
        results[seq_id] = matches
    return results
```

### Step 3: Execute Motif Search
Specify the motif and execute the search.

```python
motif = 'ATG[ACGT]{3}'  # Example motif
results = find_motifs(sequences, motif)
print(results)
```
