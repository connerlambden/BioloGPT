Análisis de la viscosidad del citoplasma en función de la concentración de macromoléculas.

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de viscosidad y concentración
concentration = np.array([0, 10, 20, 30, 40, 50])
viscosity = np.array([1, 1.5, 2, 2.5, 3, 3.5])

plt.plot(concentration, viscosity)
plt.title('Viscosidad del Citoplasma vs Concentración de Macromoléculas')
plt.xlabel('Concentración (%)')
plt.ylabel('Viscosidad (mPa.s)')
plt.grid()
plt.show()
```
