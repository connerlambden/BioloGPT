Visualizing atomic orbitals helps in understanding their shapes and electron distributions.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of points
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(1 - X**2 - Y**2)  # Example for s orbital

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.5)
ax.set_title('s Orbital')
plt.show()
```

This code generates a 3D plot of an s orbital, illustrating its spherical shape.

```python
# Additional code for p orbital visualization
Z_p = np.zeros_like(X)  # Example for p orbital
ax.plot_surface(X, Y, Z_p, alpha=0.5, color='r')
ax.set_title('p Orbital')
plt.show()
```
