This notebook will analyze the expression levels of immune checkpoint genes in Mastomys natalensis before and after anthelmintic treatment.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('immune_checkpoints_data.csv')

# Analyze the expression levels
sns.boxplot(x='treatment', y='expression_level', data=data)
plt.title('Immune Checkpoint Expression Levels')
plt.show()
```
