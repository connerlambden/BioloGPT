### Behavioral Recovery Analysis
This notebook analyzes the recovery scores from the CES and sham groups to determine the effectiveness of conditioning electrical stimulation.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Group': ['CES', 'Sham'], 'Recovery Score': [75, 45]}
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Group'], df['Recovery Score'], color=['blue', 'orange'])
plt.title('Functional Recovery Comparison')
plt.ylabel('Recovery Score')
plt.show()
```

### Discussion
The analysis shows a significant difference in recovery scores between the CES and sham groups, supporting the hypothesis that CES enhances functional recovery.
