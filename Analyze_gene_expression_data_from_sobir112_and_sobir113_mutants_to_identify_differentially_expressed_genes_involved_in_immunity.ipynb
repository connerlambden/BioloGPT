This notebook will analyze RNA-seq data from sobir1-12 and sobir1-13 mutants to identify key genes involved in their differing immune responses.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load datasets
sobir1_12_data = pd.read_csv('sobir1_12_expression.csv')
sobir1_13_data = pd.read_csv('sobir1_13_expression.csv')

# Perform differential expression analysis
merged_data = pd.merge(sobir1_12_data, sobir1_13_data, on='gene_id')
merged_data['log2_fold_change'] = np.log2(merged_data['expression_sobir1_12'] / merged_data['expression_sobir1_13'])
merged_data['p_value'] = stats.ttest_ind(sobir1_12_data['expression'], sobir1_13_data['expression'], equal_var=False).pvalue

# Filter significant genes
significant_genes = merged_data[merged_data['p_value'] < 0.05]
```

The analysis identifies genes that are significantly differentially expressed between sobir1-12 and sobir1-13, providing insights into their distinct immune responses.

```python
significant_genes.to_csv('significant_genes.csv', index=False)
```
