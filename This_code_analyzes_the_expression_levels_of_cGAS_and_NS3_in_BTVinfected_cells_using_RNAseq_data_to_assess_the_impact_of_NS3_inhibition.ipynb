### RNA-seq Data Analysis of cGAS and NS3 Expression Levels

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load RNA-seq data
rna_seq_data = pd.read_csv('path_to_rna_seq_data.csv')

# Filter for cGAS and NS3 expression levels
cgas_expression = rna_seq_data[rna_seq_data['gene'] == 'cGAS']
ns3_expression = rna_seq_data[rna_seq_data['gene'] == 'NS3']

# Plotting the expression levels
plt.figure(figsize=(10, 5))
plt.plot(cgas_expression['time'], cgas_expression['expression'], label='cGAS', color='blue')
plt.plot(ns3_expression['time'], ns3_expression['expression'], label='NS3', color='red')
plt.title('Expression Levels of cGAS and NS3 Over Time')
plt.xlabel('Time (hours post-infection)')
plt.ylabel('Expression Level')
plt.legend()
plt.show()
```

This analysis will help elucidate the relationship between NS3 inhibition and cGAS expression in BTV-infected cells.
