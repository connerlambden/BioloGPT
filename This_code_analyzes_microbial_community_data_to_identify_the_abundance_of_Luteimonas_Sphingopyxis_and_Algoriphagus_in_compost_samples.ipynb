### Microbial Community Analysis
This notebook analyzes the abundance of specific bacterial genera in compost samples.

```python
import pandas as pd

# Load microbial community data
data = pd.read_csv('compost_microbial_data.csv')

# Filter for specific genera
filtered_data = data[data['Genus'].isin(['Luteimonas', 'Sphingopyxis', 'Algoriphagus'])]

# Display the results
filtered_data.describe()
```

### Results
The analysis provides insights into the relative abundance of the target genera in different compost samples.

```python
# Visualize the abundance
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(filtered_data['Sample'], filtered_data['Abundance'])
plt.title('Abundance of Target Genera in Compost Samples')
plt.xlabel('Sample')
plt.ylabel('Abundance')
plt.show()
```
