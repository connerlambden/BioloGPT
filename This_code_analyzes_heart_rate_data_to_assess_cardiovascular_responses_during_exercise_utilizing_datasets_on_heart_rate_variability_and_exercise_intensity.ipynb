### Heart Rate Analysis During Exercise
This notebook analyzes heart rate data collected during exercise sessions to evaluate cardiovascular responses.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample heart rate data
heart_rate_data = {'Time (min)': [0, 1, 2, 3, 4], 'Heart Rate (bpm)': [84, 98, 115, 125, 140]}
df = pd.DataFrame(heart_rate_data)

# Plotting the heart rate response
plt.figure(figsize=(10, 5))
plt.plot(df['Time (min)'], df['Heart Rate (bpm)'], marker='o')
plt.title('Heart Rate Response During Exercise')
plt.xlabel('Time (minutes)')
plt.ylabel('Heart Rate (bpm)')
plt.grid()
plt.show()
```
