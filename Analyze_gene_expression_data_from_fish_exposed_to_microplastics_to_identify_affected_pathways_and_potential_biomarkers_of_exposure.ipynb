This notebook will analyze RNA-seq data from fish exposed to microplastics, focusing on differential gene expression.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('microplastics_gene_expression.csv')

# Perform differential expression analysis
# (Assuming the dataset has columns for gene names and expression levels)
results = data[data['p_value'] < 0.05]

# Visualize the results
plt.figure(figsize=(10,6))
sns.barplot(x='gene', y='log2_fold_change', data=results)
plt.title('Differential Gene Expression in Fish Exposed to Microplastics')
plt.xticks(rotation=90)
plt.show()
```

This analysis will help identify key genes affected by microplastic exposure, providing insights into the biological impact of pollution.

```python
# Save the results to a new CSV file
results.to_csv('differential_expression_results.csv', index=False)
```
