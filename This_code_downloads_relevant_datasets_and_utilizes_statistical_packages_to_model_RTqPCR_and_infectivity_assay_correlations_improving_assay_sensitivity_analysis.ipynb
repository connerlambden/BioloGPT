Detailing the download of experimental datasets and the extraction of RT-qPCR and PFU values for analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming dataset_url points to the supplementary dataset
dataset_url = 'https://example.com/dataset.csv'
df = pd.read_csv(dataset_url)

# Data filtering and linear regression modeling
X = df['log10_GC'].values.reshape(-1, 1)
y = df['log10_PFU'].values

from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X, y)

# Plotting the correlation
y_pred = model.predict(X)
plt.figure(figsize=(8,6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red', label='Linear Model')
plt.xlabel('log10 Genome Copies (GC)')
plt.ylabel('log10 PFU')
plt.title('Correlation between RT-qPCR and Infectivity Assays')
plt.legend()
plt.show()
```

This notebook provides a reproducible analysis pipeline for comparing RT-qPCR data with classical infectivity assays.

```python
# Additional code for evaluating model performance
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y, y_pred)
print(f'Mean Squared Error: {mse}')
```
