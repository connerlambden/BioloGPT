This notebook will explore chromosomal behavior during prophase I across different species, focusing on those with a diploid number of 32.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load chromosomal data
chromosomal_data = pd.read_csv('chromosomal_data.csv')

# Visualize the data
plt.figure(figsize=(10, 6))
plt.bar(chromosomal_data['species'], chromosomal_data['chromosome_count'])
plt.title('Chromosome Counts in Prophase I')
plt.xlabel('Species')
plt.ylabel('Chromosome Count')
plt.xticks(rotation=45)
plt.show()
```

This analysis provides insights into the chromosomal dynamics during prophase I, highlighting differences across species.
