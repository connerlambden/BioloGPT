### Data Analysis of Drosophila Locomotor Activity

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
activity_data = pd.read_csv('drosophila_activity.csv')

# Analyze the data
plt.figure(figsize=(10, 5))
plt.plot(activity_data['Time'], activity_data['Activity'], label='Activity Level')
plt.plot(activity_data['Time'], activity_data['Sleep'], label='Sleep Level')
plt.title('Drosophila Activity and Sleep Patterns')
plt.xlabel('Time of Day')
plt.ylabel('Level')
plt.legend()
plt.show()
```
