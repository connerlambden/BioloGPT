This notebook will analyze the correlation between maternal stress levels and the incidence of gastrointestinal issues in infants.

```python
import pandas as pd
# Load relevant datasets
maternal_stress_data = pd.read_csv('maternal_stress.csv')
infant_health_data = pd.read_csv('infant_health.csv')
```

We will merge the datasets and perform correlation analysis.

```python
merged_data = pd.merge(maternal_stress_data, infant_health_data, on='mother_id')
correlation = merged_data.corr()
```

The correlation matrix will help identify any significant relationships.

```python
print(correlation)
```
