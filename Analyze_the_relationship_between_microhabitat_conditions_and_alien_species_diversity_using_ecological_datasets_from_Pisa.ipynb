This notebook will analyze ecological data to explore how microhabitat conditions influence alien species diversity in Pisa.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('pisa_ecological_data.csv')
sns.scatterplot(data=data, x='microhabitat_condition', y='alien_species_diversity')
plt.title('Microhabitat Conditions vs Alien Species Diversity')
plt.show()
```
