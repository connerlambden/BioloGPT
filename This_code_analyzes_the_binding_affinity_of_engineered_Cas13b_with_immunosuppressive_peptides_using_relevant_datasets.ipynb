This notebook will analyze the binding affinities of various immunosuppressive peptides to engineered Cas13b.

```python
import pandas as pd
# Load relevant datasets
peptide_data = pd.read_csv('peptide_binding_affinities.csv')
# Analyze binding affinities
binding_affinities = peptide_data.groupby('peptide').mean()
```

The analysis will provide insights into which peptides have the highest binding affinities to Cas13b.

```python
binding_affinities.to_csv('average_binding_affinities.csv')
```
