This notebook will analyze aptamer sequences from SELEX data, focusing on binding affinities and sequence diversity.

```python
import pandas as pd

# Load SELEX data
selex_data = pd.read_csv('selex_data.csv')

# Analyze binding affinities
binding_affinities = selex_data.groupby('aptamer')['K_d'].mean()

# Output results
binding_affinities.to_csv('binding_affinities.csv')
```

The analysis provides insights into the binding affinities of various aptamers, helping to identify the most promising candidates.

```python
# Further analysis can include visualizing the distribution of binding affinities
import matplotlib.pyplot as plt

plt.hist(binding_affinities, bins=20)
plt.title('Distribution of Binding Affinities')
plt.xlabel('K_d (pM)')
plt.ylabel('Frequency')
plt.show()
```
