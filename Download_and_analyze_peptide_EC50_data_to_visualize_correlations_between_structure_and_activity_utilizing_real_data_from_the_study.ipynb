The following notebook cell downloads the peptide dataset and plots EC50 values for both spirochetal and cystic forms.

```python
import pandas as pd
import plotly.express as px

data = {'Peptide': ['Bor-11', 'Bor-16', 'Bor-18', 'Bor-26'], 'EC50_Spirochetal': [0.41, 0.83, 0.83, 0.83], 'EC50_Cystic': [0.41, 0.83, 0.83, 0.83]}
df = pd.DataFrame(data)
fig = px.bar(df, x='Peptide', y=['EC50_Spirochetal', 'EC50_Cystic'], barmode='group',
             labels={'value':'EC50 (µM)', 'variable':'Form'}, title='EC50 Values for Cyclic Peptides')
fig.show()
```
