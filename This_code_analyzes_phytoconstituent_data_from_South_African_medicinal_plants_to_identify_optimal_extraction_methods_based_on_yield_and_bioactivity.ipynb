# Analysis of Phytoconstituents from South African Medicinal Plants
This notebook will analyze data on phytoconstituents extracted using various methods.

```python
import pandas as pd

# Load data on extraction methods and yields
data = pd.read_csv('phytoconstituent_data.csv')

# Analyze yields by extraction method
yield_analysis = data.groupby('extraction_method')['yield'].mean()
```

## Yield Analysis
This section presents the average yield of phytoconstituents based on different extraction methods.

```python
yield_analysis.plot(kind='bar', title='Average Yield by Extraction Method')
```
