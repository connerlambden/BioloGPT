This notebook calculates the Simpson diversity index based on species abundance data.

```python
import numpy as np

def simpson_index(abundance):
    total = np.sum(abundance)
    proportions = abundance / total
    return 1 - np.sum(proportions**2)

# Example data
abundance_data = np.array([10, 20, 30])
index = simpson_index(abundance_data)
index
```

The Simpson index calculated from the example data provides insights into the diversity of the species present.

```python
print(f'Simpson Diversity Index: {index}')
```
