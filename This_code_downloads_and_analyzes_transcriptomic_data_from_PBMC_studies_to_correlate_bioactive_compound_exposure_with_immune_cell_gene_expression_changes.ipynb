Retrieve and preprocess single-cell RNA-seq datasets from the Broad Single Cell Portal related to immunomodulatory compound screens.

```python
import scanpy as sc
adata = sc.read_10x_h5('path_to_data.h5')
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)
adata.raw = adata
sc.tl.pca(adata, svd_solver='arpack')
sc.pl.pca_variance_ratio(adata, log=True)
```

Next, perform clustering and visualize the key gene expression markers responsive to bioactive compound exposure.

```python
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
sc.tl.umap(adata)
sc.tl.leiden(adata)
sc.pl.umap(adata, color=['GeneX', 'GeneY'])
```

The resulting visualizations will help identify immune cell populations most responsive to the bioactive treatments.

```python
# Export results for downstream analysis
adata.write('processed_immunomodulatory_data.h5ad')
```
