### Gene Expression Analysis of RAP2A in GBM Samples
This analysis will utilize publicly available datasets to explore the relationship between RAP2A expression and GSC markers.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv('path_to_gbm_data.csv')

# Correlation analysis
correlation = dataset[['RAP2A', 'CD133', 'SOX2', 'NESTIN']].corr()

# Heatmap visualization
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between RAP2A and GSC Markers')
plt.show()
```
