Analyze gene expression datasets from statin-treated individuals to identify differentially expressed genes associated with muscle atrophy.

```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('gene_expression_data.csv')
# Perform differential expression analysis
results = dataset[dataset['treatment'] == 'statin']
# Identify significant genes
significant_genes = results[results['p_value'] < 0.05]
```

This analysis will help identify potential biomarkers for muscle atrophy in patients on statin therapy.

```python
# Output significant genes
significant_genes.to_csv('significant_genes.csv', index=False)
```
