تحليل البيانات العشبية لتحديد النسب المثلى للأعشاب في التركيبات العلاجية.

```python
# تحميل البيانات العشبية
herb_data = {'زنجبيل': 0.1, 'كركم': 0.05, 'ريحان مقدس': 0.3, 'روزماري': 0.02}

# حساب النسب المئوية
for herb, amount in herb_data.items():
    print(f'{herb}: {amount * 100}%')
```
