RNA-seq data analysis for identifying gene expression changes.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load RNA-seq data
rna_data = pd.read_csv('rna_seq_data.csv')

# Perform differential expression analysis
results = stats.ttest_ind(rna_data['control'], rna_data['treatment'])

# Output results
results_df = pd.DataFrame({'Gene': rna_data['Gene'], 'p-value': results.pvalue})
results_df.to_csv('differential_expression_results.csv', index=False)
```
