This notebook will analyze the sorption and diffusion data of CH4 and CO2 in polysulfone membranes.

```python
import pandas as pd
import numpy as np

# Load the dataset
data = pd.read_csv('sorption_data.csv')

# Analyze the diffusion coefficients
mean_diffusion = data['diffusion_coefficient'].mean()
print(f'Mean Diffusion Coefficient: {mean_diffusion}')
```

The analysis provides insights into the average diffusion coefficients, which can be used to optimize membrane performance.

```python
# Visualize the results
import matplotlib.pyplot as plt
plt.plot(data['pressure'], data['diffusion_coefficient'])
plt.title('Diffusion Coefficient vs Pressure')
plt.xlabel('Pressure (bar)')
plt.ylabel('Diffusion Coefficient')
plt.show()
```
