This notebook will analyze the data collected from the floating leaf disk experiment, focusing on the relationship between light intensity and the number of floating disks.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('leaf_disk_data.csv')
plt.scatter(data['Light Intensity'], data['Floating Disks'])
plt.title('Light Intensity vs. Floating Disks')
plt.xlabel('Light Intensity (µmol m-2 s-1)')
plt.ylabel('Number of Floating Disks')
plt.show()
```

The scatter plot above illustrates the correlation between light intensity and the number of floating disks, providing insights into the photosynthetic efficiency under varying light conditions.
