This notebook will analyze the feeding rates of different amphipod species and their effects on nutrient cycling.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load amphipod feeding data
data = {'Species': ['Gammarus spp.', 'Tipula spp.', 'No Shredders'], 'NO3_N_High': [413.6, 510.7, 363.7], 'NH4_N_High': [20.4, 59.7, 21.4]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Species'], df['NO3_N_High'], color='blue', label='NO3-N High')
plt.bar(df['Species'], df['NH4_N_High'], color='orange', label='NH4-N High', bottom=df['NO3_N_High'])
plt.title('Nutrient Cycling Influenced by Amphipods')
plt.ylabel('Nutrient Concentration (µg/L)')
plt.legend()
plt.show()
```
