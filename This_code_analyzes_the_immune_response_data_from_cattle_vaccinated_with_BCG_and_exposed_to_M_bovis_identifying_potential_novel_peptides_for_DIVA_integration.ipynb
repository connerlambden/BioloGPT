### Data Analysis of Immune Responses in Cattle
This notebook analyzes immune response data to identify novel peptides that could enhance the DIVA cocktail.

```python
import pandas as pd

# Load immune response data
data = pd.read_csv('immune_response_data.csv')

# Analyze response patterns
response_summary = data.groupby('peptide').agg({'response_strength': 'mean'})
response_summary.sort_values(by='response_strength', ascending=False).head(10) # Top 10 peptides
```

### Summary of Findings
The analysis identifies the top peptides that elicit strong immune responses, guiding the selection for DIVA cocktail enhancement.

```python
# Visualize the top peptides
import matplotlib.pyplot as plt
plt.bar(response_summary.index, response_summary['response_strength'])
plt.title('Top Peptides by Immune Response Strength')
plt.xlabel('Peptide')
plt.ylabel('Mean Response Strength')
plt.xticks(rotation=45)
plt.show()
```
