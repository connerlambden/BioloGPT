### Genetic Diversity Analysis
This notebook analyzes genetic diversity data from various studies, focusing on ecological strategies and reproductive modes.

```python
import pandas as pd

# Load genetic diversity data
data = pd.read_csv('genetic_diversity_data.csv')

# Analyze the impact of reproductive modes on genetic diversity
selfing_data = data[data['reproductive_mode'] == 'selfing']
outcrossing_data = data[data['reproductive_mode'] == 'outcrossing']

# Calculate average genetic diversity
average_selfing_diversity = selfing_data['genetic_diversity'].mean()
average_outcrossing_diversity = outcrossing_data['genetic_diversity'].mean()

average_selfing_diversity, average_outcrossing_diversity
```

### Results
The analysis reveals differences in genetic diversity between selfing and outcrossing populations.

```python
# Display results
print(f'Average Selfing Diversity: {average_selfing_diversity}')
print(f'Average Outcrossing Diversity: {average_outcrossing_diversity}')
```
