### Serum Metabolite Analysis
This section analyzes the serum metabolite data collected from the study.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for serum metabolites
metabolite_data = {
    'Time': ['D30', 'D33', 'D36', 'D45'],
    'Control': [5.74, 8.62, 10.00, 8.64],
    'Early Weaning': [6.13, 5.70, 6.29, 6.28]
}

# Create DataFrame
df = pd.DataFrame(metabolite_data)

# Plotting
plt.figure(figsize=(10, 5))
for column in df.columns[1:]:
    plt.plot(df['Time'], df[column], marker='o', label=column)
plt.title('Serum Glucose Levels Over Time')
plt.xlabel('Days')
plt.ylabel('Glucose (mmol/L)')
plt.legend()
plt.grid()
plt.show()
```
