This notebook will analyze RNA sequencing data from urinary EVs to identify differentially expressed genes associated with immune dysregulation.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load relevant datasets
urine_ev_data = pd.read_csv('urinary_ev_data.csv')

# Perform differential expression analysis
results = stats.ttest_ind(urine_ev_data['pre_RESP'], urine_ev_data['post_RESP'])

# Output results
results_df = pd.DataFrame({'Gene': urine_ev_data['Gene'], 'p-value': results.pvalue})
results_df.to_csv('differential_expression_results.csv', index=False)
```
