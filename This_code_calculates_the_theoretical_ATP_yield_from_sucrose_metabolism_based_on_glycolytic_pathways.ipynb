### ATP Yield Calculation from Sucrose
This notebook calculates the ATP yield from the metabolism of sucrose, considering glycolysis and fermentation pathways.

```python
# Constants
sucrose_molecules = 34
atp_per_molecule = 2

# Calculate total ATP yield
atp_yield = sucrose_molecules * atp_per_molecule * 2  # 2 ATP from glucose and 2 ATP from fructose
atp_yield
```

The calculated ATP yield from 34 molecules of sucrose is displayed above.

```python
# Displaying the result
print(f'Total ATP yield from {sucrose_molecules} sucrose molecules: {atp_yield} ATP')
```
