Cette analyse explore comment la rigidité du substrat influence le comportement cellulaire en utilisant des données expérimentales.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('rigidity_data.csv')
plt.scatter(data['rigidity'], data['cell_behavior'])
plt.title('Rigidité vs Comportement Cellulaire')
plt.xlabel('Rigidité (kPa)')
plt.ylabel('Comportement Cellulaire')
plt.show()
```

Cette visualisation montre la relation entre la rigidité du substrat et le comportement cellulaire, permettant d'identifier des tendances.

```python
# Code pour analyser les données de rigidité et comportement cellulaire
import seaborn as sns
sns.boxplot(x='rigidity', y='cell_behavior', data=data)
```
