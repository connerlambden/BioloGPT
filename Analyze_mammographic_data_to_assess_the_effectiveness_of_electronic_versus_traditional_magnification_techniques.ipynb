This notebook will analyze mammographic data comparing electronic and traditional magnification techniques.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
magnification_data = pd.read_csv('magnification_data.csv')

# Analyze detectability
plt.bar(magnification_data['lesion_size'], magnification_data['detectability_magnification'], label='Magnification Film')
plt.bar(magnification_data['lesion_size'], magnification_data['detectability_digitized'], label='Digitized Image', alpha=0.7)
plt.xlabel('Lesion Size (mm)')
plt.ylabel('Detectability')
plt.title('Lesion Detectability Comparison')
plt.legend()
plt.show()
```
