This notebook will analyze socio-economic data from Nigerian mothers to identify predictors of sickle cell trait prevalence.

```python
import pandas as pd
import statsmodels.api as sm

# Load the dataset
data = pd.read_csv('sct_data.csv')

# Define the independent variables and the dependent variable
X = data[['education_level', 'income', 'region']]
y = data['sct_prevalence']

# Add a constant to the model
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the summary of the regression model
model.summary()
```

The regression analysis will provide insights into which socio-economic factors are significant predictors of SCT prevalence.

```python
# Visualize the results
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x=model.params.index, y=model.params.values)
plt.title('Predictors of SCT Prevalence')
plt.ylabel('Coefficient Value')
plt.xticks(rotation=45)
plt.show()
```
