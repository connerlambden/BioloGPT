This notebook will analyze ecological data to explore the relationships between forest microhabitat diversity, rodent populations, and TBE incidence.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
microhabitat_data = pd.read_csv('microhabitat_data.csv')
rodent_data = pd.read_csv('rodent_data.csv')
tbe_data = pd.read_csv('tbe_data.csv')

# Merge datasets
merged_data = pd.merge(microhabitat_data, rodent_data, on='year')
merged_data = pd.merge(merged_data, tbe_data, on='year')

# Analyze correlations
correlation_matrix = merged_data.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix of Ecological Factors')
plt.show()
```

This analysis will help identify significant correlations between microhabitat diversity, rodent abundance, and TBE incidence.
