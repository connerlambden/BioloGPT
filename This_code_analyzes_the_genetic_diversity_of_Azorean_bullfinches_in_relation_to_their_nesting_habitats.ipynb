### Genetic Diversity Analysis of Azorean Bullfinches
This analysis will focus on the genetic diversity of Azorean bullfinches in relation to their nesting habitats, comparing those in native versus exotic trees.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load genetic data
data = pd.read_csv('azorean_bullfinch_genetics.csv')

# Analyze genetic diversity
native_habitat = data[data['habitat'] == 'native']
exotic_habitat = data[data['habitat'] == 'exotic']

# Plotting genetic diversity
plt.figure(figsize=(10, 6))
plt.hist(native_habitat['genetic_diversity'], alpha=0.5, label='Native Habitat')
plt.hist(exotic_habitat['genetic_diversity'], alpha=0.5, label='Exotic Habitat')
plt.title('Genetic Diversity of Azorean Bullfinches')
plt.xlabel('Genetic Diversity Index')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
