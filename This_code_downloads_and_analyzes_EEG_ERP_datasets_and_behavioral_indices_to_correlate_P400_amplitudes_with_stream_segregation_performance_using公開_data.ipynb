Load EEG and behavioral dataset from the repository and perform correlation analysis between ERP features and task performance.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Replace 'data_url' with actual GitHub URL
url = 'https://github.com/acalcus/ORN_P400/raw/main/eeg_behavior_data.csv'
data = pd.read_csv(url)

# Compute correlation between P400 amplitude and segregation performance
corr, p_val = pearsonr(data['P400_amplitude'], data['segregation_performance'])
print('Correlation:', corr, 'p-value:', p_val)

plt.scatter(data['P400_amplitude'], data['segregation_performance'], color='green')
plt.title('P400 Amplitude vs Segregation Performance')
plt.xlabel('P400 Amplitude')
plt.ylabel('Segregation Performance')
plt.show()
```

The above code snippet helps test the hypothesis by plotting and analyzing the correlation between P400 ERP measures and behavioral performance, thereby offering insights into neural maturation effects.
