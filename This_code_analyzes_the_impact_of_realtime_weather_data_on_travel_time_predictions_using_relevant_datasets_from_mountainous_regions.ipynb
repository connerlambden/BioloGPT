### Data Preparation
Load datasets containing travel time and weather data for analysis.

```python
import pandas as pd
weather_data = pd.read_csv('weather_data.csv')
travel_data = pd.read_csv('travel_time_data.csv')
```

### Data Analysis
Merge datasets and analyze the correlation between weather conditions and travel times.

```python
merged_data = pd.merge(travel_data, weather_data, on='date')
correlation = merged_data.corr()
```

### Visualization
Visualize the impact of weather on travel times.

```python
import matplotlib.pyplot as plt
plt.scatter(merged_data['weather_condition'], merged_data['travel_time'])
plt.title('Impact of Weather on Travel Time')
plt.xlabel('Weather Condition')
plt.ylabel('Travel Time')
plt.show()
```
