Análisis de datos de resistencia a antibióticos en gramnegativos.

```python
import pandas as pd
# Cargar datos de resistencia a antibióticos
resistance_data = pd.read_csv('resistance_data.csv')
# Análisis de la clasificación de Ambler
ambler_classification = resistance_data.groupby('Ambler_Class').mean()
```

Resultados del análisis de la clasificación de Ambler.

```python
ambler_classification.to_csv('ambler_analysis_results.csv')
```
