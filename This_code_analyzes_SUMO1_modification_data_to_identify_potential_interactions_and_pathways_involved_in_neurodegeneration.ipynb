### SUMO-1 Modification Analysis
This notebook analyzes SUMO-1 modification data from AD mouse models to identify key interactions and pathways.

```python
import pandas as pd

# Load SUMO-1 modification data
data = pd.read_csv('sumo1_modifications.csv')

# Analyze interactions
interactions = data.groupby('protein').agg({'modification_count': 'sum'})
print(interactions)
```
