This notebook will analyze the electrostatic interactions between Cyclin F and its substrates using sequence data.

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load Cyclin F and substrate sequences
cyclin_f_sequence = 'MRYIL...'
substrate_sequences = ['E2F1_sequence', 'p130_sequence']

# Function to calculate charge distribution
def calculate_charge(sequence):
    charge = 0
    for aa in sequence:
        if aa in 'KR':  # Positive charges
            charge += 1
        elif aa in 'DE':  # Negative charges
            charge -= 1
    return charge

# Calculate charges
cyclin_f_charge = calculate_charge(cyclin_f_sequence)
substrate_charges = [calculate_charge(seq) for seq in substrate_sequences]

# Display results
print(f'Cyclin F Charge: {cyclin_f_charge}')
print(f'Substrate Charges: {substrate_charges}')
```
