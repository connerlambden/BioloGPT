This notebook will analyze RNA-seq data from Toxoplasma gondii to identify differentially expressed nuclear-mitochondrial transporters during infection.

```python
import pandas as pd
import numpy as np

# Load RNA-seq data
data = pd.read_csv('path_to_rna_seq_data.csv')

# Identify differentially expressed genes
# Assuming 'expression' column contains expression levels and 'gene' column contains gene names
threshold = 2.0
filtered_data = data[data['expression'] > threshold]

# Output the results
filtered_data.to_csv('differentially_expressed_genes.csv', index=False)
```
