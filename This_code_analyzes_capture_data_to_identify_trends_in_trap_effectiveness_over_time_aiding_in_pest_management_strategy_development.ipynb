This notebook analyzes the capture rates of Tuta absoluta based on different trap types and colors.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for analysis
data = {
    'Year': [2020, 2020, 2020, 2020, 2021, 2021, 2021, 2021],
    'Trap Type': ['Rubber septum', 'Polymer wax', 'Polymer vial', 'Control', 'Rubber septum', 'Polymer wax', 'Polymer vial', 'Control'],
    'Mean Capture': [76.0, 32.83, 10.37, 9.5, 86.17, 28.2, 11.77, 5.0]
}

df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 6))
for year in df['Year'].unique():
    subset = df[df['Year'] == year]
    plt.bar(subset['Trap Type'], subset['Mean Capture'], label=str(year))
plt.title('Capture Rates of Tuta absoluta by Trap Type')
plt.xlabel('Trap Type')
plt.ylabel('Mean Capture')
plt.legend(title='Year')
plt.show()
```

The analysis shows the effectiveness of different traps over the two years.
