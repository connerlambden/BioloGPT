### Analyzing Adhesive Properties of Biopolymers
This notebook will analyze the adhesive properties of different biopolymers based on experimental data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for analysis
data = {
    'Carbopol_934P': [3332.4, 5588.8, 7669.6, 2712.8],
    'Carbopol_974P': [3332.4, 5916.0, 8601.2, 3754.9],
    'Amount_Added': [0, 200, 400, 600]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Amount_Added'], df['Carbopol_934P'], marker='o', label='Carbopol 934P')
plt.plot(df['Amount_Added'], df['Carbopol_974P'], marker='o', label='Carbopol 974P')
plt.title('Adhesive Force of Carbopol Solutions')
plt.xlabel('Amount Added (ml)')
plt.ylabel('Adhesive Force (g)')
plt.legend()
plt.grid()
plt.show()
```

### Discussion
This analysis provides insights into how the amount of added Carbopol affects adhesive force, which is crucial for optimizing formulations.
