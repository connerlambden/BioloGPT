Load study data and visualize risk factors with a bar graph.

```python
import plotly.graph_objects as go

risk_factors = ['Age >74', 'Heart Disease', 'Hypertension >5 yrs', 'Dyslipidemia', 'Uncontrolled HTN', 'Poor Adherence', 'Diabetes']
odds_ratios = [4.24, 5.38, 12.66, 3.44, 40.03, 1.84, 3.00]

fig = go.Figure([go.Bar(x=risk_factors, y=odds_ratios, marker_color='#6A0C76')])
fig.update_layout(title='Adjusted Odds Ratios for Hypertensive Retinopathy Risk Factors',
                  xaxis_title='Risk Factors',
                  yaxis_title='Adjusted Odds Ratio (AOR)')
fig.show()
```

This visualization aids in quickly understanding which factors most significantly contribute to hypertensive retinopathy risk.
