### Step 1: Import Libraries
Import necessary libraries for data analysis.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2: Load Data
Load gene expression data from Saccharina japonica.

```python
data = pd.read_csv('saccharina_japonica_expression.csv')
print(data.head())
```

### Step 3: Analyze Differential Expression
Identify differentially expressed genes under stress conditions.

```python
# Assuming 'stress_condition' is a column indicating stress exposure
stress_genes = data[data['stress_condition'] == 'yes']
print(stress_genes.describe())
```

### Step 4: Visualize Results
Create visualizations to represent the expression levels of adaptive genes.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='gene', y='expression_level', data=stress_genes)
plt.title('Gene Expression Levels Under Stress')
plt.xticks(rotation=90)
plt.show()
```
