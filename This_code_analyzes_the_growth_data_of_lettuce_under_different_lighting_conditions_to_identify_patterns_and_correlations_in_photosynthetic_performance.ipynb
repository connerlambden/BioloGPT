### Data Analysis of Lettuce Growth Under Dynamic Lighting
This notebook will analyze the growth data collected from the study on lettuce under various lighting regimes.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Lighting Regime': ['Constant Light', 'Split-Night', 'Price-Based'], 'Fresh Weight (g)': [3.5, 3.4, 3.6]}
df = pd.DataFrame(data)

plt.bar(df['Lighting Regime'], df['Fresh Weight (g)'])
plt.title('Lettuce Growth Under Different Lighting Regimes')
plt.xlabel('Lighting Regime')
plt.ylabel('Fresh Weight (g)')
plt.show()
```
