This notebook demonstrates the process of retrieving and analyzing GPCR expression datasets from metastatic tumor samples.

```python
import pandas as pd
url = 'https://example.com/gpcr_metastasis_data.csv'
df = pd.read_csv(url)
df_summary = df.groupby('GPCR').agg({'expression':'mean'})
print(df_summary.head())
```

The subsequent analysis involves plotting average expression levels to visually compare target profiles.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
df_summary['expression'].plot(kind='bar', color='#6A0C76')
plt.title('Average GPCR Expression in Metastatic Tumors')
plt.xlabel('GPCR')
plt.ylabel('Mean Expression')
plt.show()
```
