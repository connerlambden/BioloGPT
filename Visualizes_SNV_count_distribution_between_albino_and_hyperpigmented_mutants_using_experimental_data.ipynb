In this notebook, we load SNV count data for alb and hyp mutants and create a comparative plot using Plotly to illustrate mutational differences.

```python
import plotly.express as px
import pandas as pd

# Sample data representing SNV counts for demonstration
# In practice, load real experimental SNV data

data = {'Mutant_Type': ['albino', 'albino', 'hypo', 'hypo', 'hypo'],
        'SNV_Count': [5, 7, 15, 20, 18]}
df = pd.DataFrame(data)

fig = px.box(df, x='Mutant_Type', y='SNV_Count', color='Mutant_Type',
             title='SNV Distribution in Albino vs Hyper-pigmented Mutants')
fig.show()
```

The above code block creates a boxplot that visualizes the SNV count differences between albino and hyper-pigmented mutants, helping to elucidate the underlying genetic diversification.
