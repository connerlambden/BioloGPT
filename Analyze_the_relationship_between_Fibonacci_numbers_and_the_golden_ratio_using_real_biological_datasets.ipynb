This notebook will explore the connection between Fibonacci numbers and the golden ratio in biological contexts.

```python
import numpy as np
import matplotlib.pyplot as plt

# Fibonacci sequence generation
fib = [0, 1]
for i in range(2, 20):
    fib.append(fib[-1] + fib[-2])

# Calculate golden ratio approximations
golden_ratios = [fib[i+1] / fib[i] for i in range(1, len(fib)-1)]

# Plotting
plt.plot(golden_ratios, marker='o')
plt.title('Approximations of the Golden Ratio from Fibonacci Sequence')
plt.xlabel('n')
plt.ylabel('F(n+1)/F(n)')
plt.axhline(y=(1 + np.sqrt(5)) / 2, color='r', linestyle='--')
plt.show()
```

This analysis visualizes how the ratios of successive Fibonacci numbers converge to the golden ratio.
