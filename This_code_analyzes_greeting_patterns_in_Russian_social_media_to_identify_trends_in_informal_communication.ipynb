### Data Collection
Collect data from Russian social media platforms to analyze greeting usage.

```python
import pandas as pd
# Load data
social_media_data = pd.read_csv('greeting_data.csv')
```

### Data Analysis
Analyze the frequency of 'Привет' in different contexts.

```python
greeting_counts = social_media_data['greeting'].value_counts()
```

### Visualization
Visualize the results to understand trends.

```python
import matplotlib.pyplot as plt
plt.bar(greeting_counts.index, greeting_counts.values)
plt.title('Usage of Greetings in Social Media')
plt.xlabel('Greeting')
plt.ylabel('Frequency')
plt.show()
```
