### Analysis of Antibiotic Resistance in ON Pathogens

```python
import pandas as pd

# Load antibiotic resistance data
data = pd.read_csv('antibiotic_resistance_on.csv')

# Analyze resistance patterns
resistance_summary = data.groupby('Pathogen').agg({'Resistance_Rate': 'mean'})
print(resistance_summary)
```
