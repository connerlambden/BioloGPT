### Patient Data Analysis for Zr-89 Imaging Outcomes

```python
import pandas as pd

# Load patient data
patient_data = pd.read_csv('patient_data.csv')

# Analyze correlation between Zr-89 exposure and imaging outcomes
correlation = patient_data.corr()['Zr-89_exposure']['imaging_outcome']
print(f'Correlation between Zr-89 exposure and imaging outcome: {correlation}')
```

### Discussion
This analysis will help identify any significant relationships between Zr-89 exposure levels and patient imaging outcomes.
