### Lensless Imaging Data Analysis
This notebook analyzes imaging data obtained from lensless microscopy techniques.

```python
import pandas as pd
import numpy as np

# Load imaging data
data = pd.read_csv('imaging_data.csv')

# Analyze cellular structures
cell_count = data['cell_id'].nunique()
mean_size = data['size'].mean()

print(f'Total cells detected: {cell_count}')
print(f'Mean cell size: {mean_size}')
```

### Results
The analysis provides insights into the cellular structures observed in the lensless imaging data.

```python
# Visualize results
import matplotlib.pyplot as plt

plt.hist(data['size'], bins=30)
plt.title('Cell Size Distribution')
plt.xlabel('Size (µm)')
plt.ylabel('Frequency')
plt.show()
```
