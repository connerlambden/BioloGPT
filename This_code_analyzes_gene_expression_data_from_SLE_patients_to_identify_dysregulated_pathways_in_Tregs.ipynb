This notebook will analyze RNA-seq data from SLE patients to identify key dysregulated genes in Tregs.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load SLE gene expression data
gene_data = pd.read_csv('path_to_sle_gene_data.csv')

# Analyze Treg-specific gene expression
# Filter for Treg-related genes
treg_genes = gene_data[gene_data['Gene'].isin(['FOXP3', 'CTLA4', 'IL2RA'])]

# Visualize the expression levels
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gene', y='Expression', data=treg_genes)
plt.title('Treg Gene Expression in SLE Patients')
plt.show()
```

This analysis provides insights into the expression levels of key Treg genes in SLE patients.

```python
# Save the results for further analysis
treg_genes.to_csv('treg_gene_expression_analysis.csv', index=False)
```
