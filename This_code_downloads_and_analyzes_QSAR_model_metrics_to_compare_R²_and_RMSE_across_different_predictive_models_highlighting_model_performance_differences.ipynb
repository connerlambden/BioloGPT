Overview: This notebook loads the performance metrics for MLR, RF, and ANN models and visualizes them using bar plots for R² and RMSE.

```python
import pandas as pd
import plotly.express as px

# Create a dataframe with model performance metrics
data = {'Model': ['MLR', 'RF', 'ANN'],
        'R2': [0.83, 0.82, 0.92],
        'RMSE': [0.27, 0.18, 0.09]}
df = pd.DataFrame(data)

# Plot the R2 values
fig_r2 = px.bar(df, x='Model', y='R2', color='Model', title='QSAR Model Comparison: R² Values')
fig_r2.show()

# Plot the RMSE values
fig_rmse = px.bar(df, x='Model', y='RMSE', color='Model', title='QSAR Model Comparison: RMSE Values')
fig_rmse.show()
```

The plots above clearly illustrate that the ANN model has the highest R² and the lowest RMSE, indicating its superior performance in predicting inhibitory activities.
