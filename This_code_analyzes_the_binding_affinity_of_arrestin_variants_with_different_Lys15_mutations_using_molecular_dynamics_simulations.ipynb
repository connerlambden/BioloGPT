This notebook will simulate the binding interactions of arrestin variants with phosphorylated rhodopsin.

```python
# Import necessary libraries
import mdtraj as md
import numpy as np

# Load the arrestin and rhodopsin structures
arrestin = md.load('arrestin.pdb')
rhodopsin = md.load('rhodopsin.pdb')
```

The following code will calculate the binding free energy between arrestin and rhodopsin.

```python
# Calculate binding free energy
binding_energy = md.compute_binding_energy(arrestin, rhodopsin)
print(f'Binding Energy: {binding_energy}')
```

This analysis will help elucidate the role of Lys-15 in the binding dynamics.

```python
# Further analysis and visualization
md.view(arrestin, rhodopsin)
```
