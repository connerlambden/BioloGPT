### Iceberg Lettuce Fat Adsorption Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for fat adsorption
data = {'Fat Type': ['Olive Oil', 'Butter', 'Lamb Fat', 'Lard'], 'Adsorption (g/100g IL)': [18, 19, 43, 33]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Fat Type'], df['Adsorption (g/100g IL)'], color='green')
plt.title('Fat Adsorption by Iceberg Lettuce')
plt.xlabel('Fat Type')
plt.ylabel('Adsorption (g/100g IL)')
plt.show()
```
