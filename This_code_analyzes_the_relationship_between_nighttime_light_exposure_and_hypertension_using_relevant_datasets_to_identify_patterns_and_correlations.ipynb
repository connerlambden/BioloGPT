### Data Analysis of Nighttime Light Exposure and Hypertension
This notebook will analyze the correlation between nighttime light exposure and hypertension using datasets from recent studies.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
nle_data = pd.read_csv('nighttime_light_exposure.csv')
hypertension_data = pd.read_csv('hypertension_data.csv')

# Merge datasets on common identifiers
merged_data = pd.merge(nle_data, hypertension_data, on='location_id')

# Analyze correlation
correlation = merged_data.corr()['hypertension_rate']['nighttime_light_exposure']
print(f'Correlation between NLE and Hypertension: {correlation}')
```

### Results
The analysis will provide insights into how nighttime light exposure correlates with hypertension rates across different populations.

```python
# Visualize the relationship
sns.scatterplot(data=merged_data, x='nighttime_light_exposure', y='hypertension_rate')
plt.title('Nighttime Light Exposure vs Hypertension Rate')
plt.xlabel('Nighttime Light Exposure (Lux)')
plt.ylabel('Hypertension Rate (%)')
plt.show();
```
