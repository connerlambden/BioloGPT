This notebook will simulate the interactions of adenine, guanine, cytosine, and thymine in a DNA model.

```python
# Import necessary libraries
import mdtraj as md
import numpy as np

# Load DNA structure
structure = md.load('dna.pdb')

# Analyze interactions
interactions = md.compute_contacts(structure) # Compute contacts between nucleobases.
```

The results will provide insights into the stability and interactions of different nucleobases.

```python
# Visualize interactions
import matplotlib.pyplot as plt
plt.hist(interactions)
plt.title('Nucleobase Interactions')
plt.xlabel('Distance (nm)')
plt.ylabel('Frequency')
plt.show()
```
