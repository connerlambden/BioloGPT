### Analysis of Filler Types in Epoxy Composites
This notebook will analyze the mechanical properties of epoxy composites reinforced with various fillers.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for analysis
filler_data = {
    'Filler Type': ['CNC', 'CNT', 'Coral'],
    'Tensile Strength (MPa)': [70, 55, 45],
    'Flexural Strength (MPa)': [103, 114, 78]
}

# Create DataFrame
filler_df = pd.DataFrame(filler_data)

# Plotting the data
filler_df.plot(x='Filler Type', kind='bar', title='Mechanical Properties of Epoxy Composites', ylabel='Strength (MPa)')
plt.show()
```
