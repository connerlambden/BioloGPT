### Data Analysis of Mediterranean Diet and Cognitive Function
This notebook will analyze the relationship between adherence to the Mediterranean diet and cognitive outcomes using relevant datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
med_diet_data = pd.read_csv('mediterranean_diet_data.csv')
cognitive_data = pd.read_csv('cognitive_outcomes.csv')

# Merge datasets on participant ID
merged_data = pd.merge(med_diet_data, cognitive_data, on='participant_id')

# Analyze correlation between diet adherence and cognitive scores
correlation = merged_data['diet_adherence'].corr(merged_data['cognitive_score'])

# Plot results
plt.scatter(merged_data['diet_adherence'], merged_data['cognitive_score'])
plt.title('Diet Adherence vs Cognitive Score')
plt.xlabel('Diet Adherence')
plt.ylabel('Cognitive Score')
plt.show()
```
