### Step 1: Load Libraries and Create Dataset
We prepare a dataset comparing SANS ambages, Mashtree, and Phylonium runtime.

```python
import pandas as pd
import plotly.express as px

# Create example runtime dataset (in minutes)
data = {'Tool': ['SANS ambages', 'Mashtree', 'Phylonium'], 'Runtime': [2, 4, 3]}
df = pd.DataFrame(data)

# Generate bar chart
fig = px.bar(df, x='Tool', y='Runtime', title='Runtime Comparison of Phylogenetic Tools', color='Tool')
fig.show()

print(df.head())
```

### Step 2: Compute Performance Improvement
Calculate the percentage improvement in runtime using SANS ambages compared to Mashtree.

```python
import numpy as np
runtime_sans = 2
runtime_mashtree = 4
improvement = ((runtime_mashtree - runtime_sans) / runtime_mashtree) * 100
print('Runtime improvement:', improvement, '%')
```
