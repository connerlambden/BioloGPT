### Biocompatibility Analysis of Nanofibers
This notebook will analyze the biocompatibility data of rPET/Cu-Alg nanofibers in comparison to other antimicrobial nanofiber composites. We will visualize the data to identify trends.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for biocompatibility scores
nanofiber_data = {'Composite': ['rPET/Cu-Alg', 'Chitosan', 'Silver Nanoparticles'], 'Biocompatibility Score': [85, 78, 80]}
df = pd.DataFrame(nanofiber_data)

# Plotting the data
plt.bar(df['Composite'], df['Biocompatibility Score'], color=['blue', 'green', 'red'])
plt.title('Biocompatibility Scores of Antimicrobial Nanofibers')
plt.xlabel('Nanofiber Composite')
plt.ylabel('Biocompatibility Score')
plt.show()
```
