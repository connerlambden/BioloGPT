### Simulating ABO Blood Type Inheritance
This notebook will simulate the inheritance of blood types based on parental genotypes. It will help visualize the probabilities of different blood types in offspring.

```python
# Define parental genotypes
father_genotype = ['I_A', 'i']  # A blood type with O allele
mother_genotype = ['I_B', 'i']  # B blood type with O allele

# Function to simulate offspring blood types
import random

def simulate_offspring(father, mother):
    offspring = []
    for _ in range(1000):  # Simulate 1000 offspring
        child = random.choice(father) + random.choice(mother)
        offspring.append(child)
    return offspring

# Simulate and count blood types
offspring = simulate_offspring(father_genotype, mother_genotype)
from collections import Counter
blood_type_counts = Counter(offspring)

blood_type_counts
```

### Results
The simulation results will show the distribution of blood types among the offspring based on the defined parental genotypes.

```python
# Display results
import matplotlib.pyplot as plt

labels = blood_type_counts.keys()
values = blood_type_counts.values()

plt.bar(labels, values)
plt.title('Blood Type Distribution in Offspring')
plt.xlabel('Blood Type')
plt.ylabel('Number of Offspring')
plt.show()
```
