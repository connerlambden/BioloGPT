This notebook loads the metabolomics dataset from firefighters and applies linear mixed effects modeling to assess differential metabolite expression by ethnicity.

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import mixedlm

data = pd.read_csv('firefighters_metabolomics.csv')
model = mixedlm('metabolite ~ ethnicity', data, groups=data['subject'])
result = model.fit()
print(result.summary())
```
