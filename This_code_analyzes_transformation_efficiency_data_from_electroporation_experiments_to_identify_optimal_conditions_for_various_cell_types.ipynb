### Electroporation Efficiency Analysis
This notebook analyzes transformation efficiency data to determine optimal electroporation conditions.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('transformation_efficiency.csv')
plt.figure(figsize=(10,6))
plt.bar(data['Cell_Type'], data['Efficiency'])
plt.title('Transformation Efficiency by Cell Type')
plt.xlabel('Cell Type')
plt.ylabel('Efficiency (colonies/µg)')
plt.show()
```
