This notebook will analyze resting-state fMRI data to evaluate the impact of early ART on functional connectivity in children with perinatally acquired HIV.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load RS-fMRI data
rs_fmri_data = pd.read_csv('rs_fmri_data.csv')

# Analyze connectivity changes
connectivity_changes = rs_fmri_data.groupby('time_point').mean()
plt.plot(connectivity_changes)
plt.title('Functional Connectivity Changes Over Time')
plt.xlabel('Time Point')
plt.ylabel('Connectivity Measure')
plt.show()
```
