### Data Analysis of Biodiversity Metrics
This notebook analyzes the biodiversity metrics collected from Knepp Estate and Boothby Wildland.

```python
import pandas as pd
import plotly.express as px

# Data for analysis
data = {
    'Year': [2022, 2022, 2023, 2024],
    'Location': ['Knepp', 'Boothby', 'Boothby', 'Knepp'],
    'Species Richness': [102, 78, 186, 162],
    'Faiths PD': [908, 391, 875, None]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
fig = px.bar(df, x='Year', y='Species Richness', color='Location', barmode='group', title='Species Richness Comparison')
fig.show()
```
