Step-by-step analysis of exome sequencing data, variant filtering, and correlation with sperm motility indices.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load variant data
variants = pd.read_csv('dnaH12_variants.csv')
# Filter potential pathogenic variants
pathogenic = variants[variants['cadd_score'] > 10]
# Plot distribution of CADD scores
plt.hist(pathogenic['cadd_score'], bins=20, color='#6A0C76')
plt.xlabel('CADD Score')
plt.ylabel('Frequency')
plt.title('Distribution of CADD Scores for DNAH12 Variants')
plt.show()
```
