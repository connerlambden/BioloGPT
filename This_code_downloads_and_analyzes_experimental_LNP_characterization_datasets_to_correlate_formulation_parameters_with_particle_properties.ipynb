```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load experimental dataset
# df = pd.read_csv('lnp_data.csv')
# For demonstration, create a sample dataframe
sample_data = {'pH': [4, 5, 6, 4, 5, 6], 'Particle_Size': [84, 90, 106, 83, 91, 108], 'EE': [91, 87, 94, 90, 88, 95], 'Zeta_Potential': [0.87, -0.58, -1.2, 0.9, -0.6, -1.3]}
df = pd.DataFrame(sample_data)

# Plot correlation between pH and Particle Size
sns.scatterplot(data=df, x='pH', y='Particle_Size')
plt.title('Correlation between mRNA aqueous solution pH and Particle Size')
plt.xlabel('pH')
plt.ylabel('Particle Size (nm)')
plt.show()
```
