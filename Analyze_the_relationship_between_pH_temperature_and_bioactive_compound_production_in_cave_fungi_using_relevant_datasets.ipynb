This notebook will explore the effects of pH and temperature on the production of bioactive compounds in cave fungi.

```python
import pandas as pd

# Load relevant datasets
datasets = pd.read_csv('cave_fungi_data.csv')

# Analyze the relationship between pH, temperature, and compound production
results = datasets.groupby(['pH', 'Temperature']).agg({'Compound_Production': 'mean'})
results.to_csv('analyzed_results.csv')
```

The analysis will provide insights into how environmental factors influence fungal metabolism.

```python
# Visualize the results
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(results['pH'], results['Temperature'], c=results['Compound_Production'], cmap='viridis')
plt.colorbar(label='Bioactive Compound Production')
plt.xlabel('pH')
plt.ylabel('Temperature')
plt.title('Effect of pH and Temperature on Bioactive Compound Production')
plt.show()
```
