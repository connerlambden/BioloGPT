This notebook will analyze pharmacokinetic data from clinical studies on Y90 microspheres.

```python
import pandas as pd
# Load pharmacokinetic data
pk_data = pd.read_csv('y90_pharmacokinetics.csv')
# Analyze data for optimal dosing strategies.
```

The analysis will focus on correlating dosing strategies with patient outcomes.

```python
# Example analysis code
optimal_dose = pk_data.groupby('liver_function').mean()['dose']
```

This will help identify the best dosing strategies for different patient profiles.

```python
# Output optimal dosing strategies
print(optimal_dose)
```
