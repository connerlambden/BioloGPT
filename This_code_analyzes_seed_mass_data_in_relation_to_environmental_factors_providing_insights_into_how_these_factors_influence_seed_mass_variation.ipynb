### Seed Mass Analysis
This notebook analyzes the relationship between seed mass and environmental factors such as precipitation and temperature.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load seed mass data
seed_data = pd.read_csv('seed_mass_data.csv')

# Analyze the relationship between precipitation and seed mass
plt.scatter(seed_data['precipitation'], seed_data['seed_mass'])
plt.title('Seed Mass vs. Precipitation')
plt.xlabel('Precipitation (mm)')
plt.ylabel('Seed Mass (mg)')
plt.show()
```
