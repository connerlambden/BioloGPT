Ova analiza koristi podatke o krvnom tlaku i razini fizičke aktivnosti za razvoj personaliziranih preporuka.

```python
import pandas as pd

# Učitavanje podataka o krvnom tlaku i fizičkoj aktivnosti
data = pd.read_csv('blood_pressure_activity_data.csv')

# Analiza podataka
results = data.groupby('activity_level').agg({'blood_pressure': 'mean'})
results.to_csv('personalized_recommendations.csv')
```

Ova analiza omogućava razvoj personaliziranih preporuka na temelju razine fizičke aktivnosti i krvnog tlaka.

```python
# Vizualizacija rezultata
import matplotlib.pyplot as plt
plt.bar(results.index, results['blood_pressure'])
plt.title('Prosječni krvni tlak prema razini fizičke aktivnosti')
plt.xlabel('Razina fizičke aktivnosti')
plt.ylabel('Prosječni krvni tlak')
plt.show()
```
