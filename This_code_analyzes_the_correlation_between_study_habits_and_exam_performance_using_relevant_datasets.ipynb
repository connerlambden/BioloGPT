### Analyzing Study Habits and Exam Performance

```python
import pandas as pd

# Load relevant datasets
dataset = pd.read_csv('study_habits_data.csv')

# Analyze correlation between study hours and exam scores
correlation = dataset['study_hours'].corr(dataset['exam_scores'])
print(f'Correlation between study hours and exam scores: {correlation}')
```

### Discussion
This analysis will help identify effective study strategies that correlate with higher exam scores.

```python
# Further analysis can be conducted to visualize the data
import matplotlib.pyplot as plt
plt.scatter(dataset['study_hours'], dataset['exam_scores'])
plt.title('Study Hours vs Exam Scores')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.show()
```
