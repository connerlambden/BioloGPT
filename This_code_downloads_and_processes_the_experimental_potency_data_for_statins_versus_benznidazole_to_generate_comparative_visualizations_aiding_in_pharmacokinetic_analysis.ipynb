The dataset includes EC50 values for statins and benznidazole against T. cruzi amastigotes and epimastigotes. This notebook will create visualizations to compare drug potencies.

```python
import pandas as pd
import plotly.express as px

# Define the data
data = {'Drug': ['Fluvastatin', 'Pravastatin', 'Simvastatin', 'Benznidazole'],
        'EC50_amastigotes': [10.3, 42, 5.7, 0.5],
        'EC50_epimastigotes': [45, 47, 27, 4.6]}
df = pd.DataFrame(data)

# Create group bar plot
fig = px.bar(df, x='Drug', y=['EC50_amastigotes', 'EC50_epimastigotes'], barmode='group',
             labels={'value':'EC50 (µM)', 'variable':'Cell Type'},
             title='In Vitro EC50 of Drugs Against T. cruzi')
fig.show()
```

The above code provides a clear visualization comparing the EC50 values for each drug, which is essential for assessing their potency in vitro.
