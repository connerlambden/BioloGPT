### Analysis of PBS Microcapsules Degradation and Adhesive Performance
This notebook will analyze the degradation kinetics of PBS microcapsules and their impact on adhesive performance using relevant datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for isocyanate loading over time
data = {
    'Time': ['1 Day', '1 Month', '2 Months', '3 Months'],
    'MCs-Monomer': [67, 56, 42, 10],
    'MCs-Trimer': [70, 64, 10, 10],
    'MCs-Polymer': [69, 54, 14, 10]
}

df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 6))
for column in df.columns[1:]:
    plt.plot(df['Time'], df[column], marker='o', label=column)
plt.title('Isocyanate Loading Over Time in PBS Microcapsules')
plt.xlabel('Time')
plt.ylabel('Isocyanate Loading (wt%)')
plt.legend()
plt.grid()
plt.show()
```
