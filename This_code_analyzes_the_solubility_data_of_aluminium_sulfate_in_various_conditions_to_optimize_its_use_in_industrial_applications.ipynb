### Solubility Analysis of Aluminium Sulfate
This notebook will analyze the solubility of aluminium sulfate under different temperature conditions.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for solubility
data = {'Temperature (°C)': [0, 25, 50, 75, 100], 'Solubility (g/L)': [1.5, 5.0, 10.0, 15.0, 20.0]}
df = pd.DataFrame(data)

plt.plot(df['Temperature (°C)'], df['Solubility (g/L)'], marker='o')
plt.title('Solubility of Aluminium Sulfate at Different Temperatures')
plt.xlabel('Temperature (°C)')
plt.ylabel('Solubility (g/L)')
plt.grid()
plt.show()
```

### Discussion
This analysis helps in understanding how temperature affects the solubility of aluminium sulfate, which is crucial for its application in water treatment.
