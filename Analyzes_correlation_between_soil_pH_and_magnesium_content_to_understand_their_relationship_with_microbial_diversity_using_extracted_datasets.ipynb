### Soil pH and Magnesium Content Correlation Analysis
This analysis explores the relationship between soil pH and magnesium content across different sample locations.

```python
import pandas as pd
import plotly.express as px

# Sample Data
data = {
    'Sample ID': [1, 2, 3],
    'Location': ['Theetukal', 'Ooty', 'Kothagiri'],
    'Latitude': ['11°24′08.0"N', '11°25′28.1"N', '11°27′38.2"N'],
    'Longitude': ['76°39′50.6"E', '76°43′23.5"E', '76°53′21.9"E'],
    'pH': [5.24, 5.46, 5.68],
    'Mg Content': [10.56, 12.34, 15.78]
}

df = pd.DataFrame(data)

# Convert Mg Content to numeric
# Assuming pH is already numeric

# Plotting pH vs. Mg Content
fig = px.scatter(df, x='pH', y='Mg Content', text='Location', title='Correlation between Soil pH and Magnesium Content', labels={'pH': 'Soil pH', 'Mg Content': 'Magnesium Content (mg kg^-1)'})
fig.update_traces(marker=dict(size=12,
                                  line=dict(width=2,
                                            color='DarkSlateGrey')),
                  selector=dict(mode='markers+text'))
fig.show()
```

### Interpretation
The scatter plot illustrates a positive correlation between soil pH and magnesium content, indicating that as pH increases, magnesium availability also increases, potentially enhancing microbial diversity.
