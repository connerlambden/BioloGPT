This notebook will analyze RNA-seq data to explore the expression of ly6g in different immune cell populations.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load RNA-seq data
rna_data = pd.read_csv('path_to_rna_seq_data.csv')

# Filter for ly6g expression
ly6g_data = rna_data[rna_data['gene'] == 'ly6g']

# Plot expression levels
plt.figure(figsize=(10,6))
sns.boxplot(x='condition', y='expression', data=ly6g_data)
plt.title('Ly6G Expression Across Conditions')
plt.show()
```

This analysis will help visualize how ly6g expression varies across different immune conditions.
