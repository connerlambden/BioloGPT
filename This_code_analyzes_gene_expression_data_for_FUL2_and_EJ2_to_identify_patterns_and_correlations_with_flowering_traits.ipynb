### Gene Expression Analysis of FUL-2 and EJ-2

```python
import pandas as pd

# Load gene expression data
gene_data = pd.read_csv('gene_expression_data.csv')

# Analyze expression levels
ful2_expression = gene_data[gene_data['gene'] == 'FUL-2']
ej2_expression = gene_data[gene_data['gene'] == 'EJ-2']

# Correlation analysis
correlation = ful2_expression['expression'].corr(ej2_expression['expression'])
print('Correlation between FUL-2 and EJ-2:', correlation)
```
