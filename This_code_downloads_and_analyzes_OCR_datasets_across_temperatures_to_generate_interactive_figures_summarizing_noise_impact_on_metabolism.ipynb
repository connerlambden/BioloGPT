The notebook loads the OCR dataset for juvenile large yellow croaker exposed to boat noise across 18°C, 25°C, and 30°C. It then computes summary statistics and visualizes the OCR differences.

```python
import pandas as pd
import plotly.express as px

df = pd.DataFrame({
    'Temperature': ['18 °C', '25 °C', '30 °C'],
    'Control_Mass_Specific': [0.20, 0.34, 0.45],
    'Noise_Mass_Specific': [0.33, 0.46, 0.58]
})

fig = px.bar(df, x='Temperature', y=['Control_Mass_Specific', 'Noise_Mass_Specific'], barmode='group',
             labels={'value': 'OCR (mg•g⁻¹•h⁻¹)', 'variable': 'Condition'},
             title='Mass-Specific Oxygen Consumption Rates')
fig.show()
```

This analysis supports visual exploration of the dataset, emphasizing the increased metabolic load under noise exposure.
