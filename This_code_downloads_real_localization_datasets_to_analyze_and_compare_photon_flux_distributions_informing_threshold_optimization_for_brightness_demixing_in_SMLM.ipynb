The following notebook steps through data loading, histogram visualization, and statistical analysis of photon flux distributions from brightness demixing experiments.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Downloading the localization dataset
url = 'https://doi.org/10.5281/zenodo.14505755'
df = pd.read_csv(url)

# Plot histogram of photon counts
df['photon_flux'].hist(bins=50)
plt.title('Photon Flux Distribution')
plt.xlabel('Photon Count')
plt.ylabel('Frequency')
plt.show()
```

The code below further analyzes localization data by comparing photon flux distributions across detected fluorophores.

```python
import seaborn as sns
sns.boxplot(data=df, x='fluorophore', y='photon_flux')
plt.title('Photon Flux by Fluorophore')
plt.show()

# Summary Statistics
print(df.describe())
```
