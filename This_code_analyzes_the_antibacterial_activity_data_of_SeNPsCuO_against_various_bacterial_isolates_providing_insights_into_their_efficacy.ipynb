### Antibacterial Activity Analysis of SeNPs/CuO
This notebook analyzes the MIC and MBC data for the SeNPs/CuO nanocomposite against different bacterial isolates.

```python
import pandas as pd

# Data for analysis
data = {
    'Bacterial Isolate': ['Staph. aureus', 'Staph. hominis', 'E. faecalis', 'E. coli', 'K. pneumoniae', 'P. aeruginosa', 'A. baumannii'],
    'MIC (µg/ml)': [6.25, 3.12, 12.5, 25, 25, 50, 50],
    'MBC (µg/ml)': [50, 25, 100, 50, 50, 100, 50]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate MBC/MIC ratio
df['MBC/MIC Ratio'] = df['MBC (µg/ml)'] / df['MIC (µg/ml)']

# Display the DataFrame
print(df)
```
