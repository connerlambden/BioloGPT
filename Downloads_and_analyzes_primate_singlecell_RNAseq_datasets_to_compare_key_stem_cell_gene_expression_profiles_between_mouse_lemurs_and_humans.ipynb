This notebook section loads the single-cell RNAseq data for Microcebus murinus, filters low-quality cells, and performs PCA to identify key gene expression patterns.

```python
import scanpy as sc
adata = sc.read_h5ad('path/to/mouselemur_data.h5ad')
sc.pp.filter_cells(adata, min_genes=200)
sc.tl.pca(adata)
sc.pl.pca(adata, color='cell_type')
```

Next, we visualize the highest expressed genes to pinpoint markers relevant to quiescent stem cells.

```python
sc.pl.highest_expr_genes(adata, n_top=20)
```
