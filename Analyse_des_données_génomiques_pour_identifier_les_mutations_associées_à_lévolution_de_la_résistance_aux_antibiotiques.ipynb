Analyse des mutations dans les génomes de bactéries résistantes.

```python
import pandas as pd
# Charger les données de mutations
mutations = pd.read_csv('mutations.csv')
# Analyser les fréquences des mutations
frequencies = mutations['mutation'].value_counts()
```

Visualisation des résultats pour identifier les mutations les plus fréquentes.

```python
import matplotlib.pyplot as plt
frequencies.plot(kind='bar')
plt.title('Fréquence des mutations')
plt.xlabel('Mutations')
plt.ylabel('Fréquence')
plt.show()
```
