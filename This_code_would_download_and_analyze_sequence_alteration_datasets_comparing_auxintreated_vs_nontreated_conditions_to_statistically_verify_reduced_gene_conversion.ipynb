This notebook downloads the sequencing data from the paper's supplementary data, computes mutation frequencies, and plots the comparative histograms.

```python
import pandas as pd
import plotly.express as px

# Load dataset from provided source
data = pd.read_csv('supplementary_data.csv')

# Compute alteration frequencies
freq = data.groupby('condition')['mutation_count'].sum().reset_index()
fig = px.bar(freq, x='condition', y='mutation_count', title='Gene Conversion Frequency Comparison')
fig.show()
```
