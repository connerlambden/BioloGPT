This notebook will analyze blood pressure data to explore the relationship between oscillometric measurements and mean arterial pressure.

```python
import pandas as pd
import numpy as np

# Load blood pressure dataset
dataset = pd.read_csv('blood_pressure_data.csv')

# Calculate MAP from SP and DP
# MAP = DP + 1/3(SP - DP)
dataset['MAP'] = dataset['DP'] + (1/3) * (dataset['SP'] - dataset['DP'])

# Analyze correlation between oscillometric measurements and MAP
correlation = dataset[['oscillation_amplitude', 'MAP']].corr()
print(correlation)
```
