Analyse des données transcriptomiques pour identifier les gènes régulateurs de la sénescence dans les cultures.

```python
import pandas as pd

# Charger les données transcriptomiques
transcript_data = pd.read_csv('data/transcriptomics.csv')

# Analyser les gènes régulateurs
regulatory_genes = transcript_data[transcript_data['type'] == 'regulatory']
print(regulatory_genes)
```

Cette analyse permet d'identifier les gènes clés impliqués dans la régulation de la sénescence, ce qui peut informer les stratégies d'amélioration des cultures.

```python
# Visualiser les gènes régulateurs
import matplotlib.pyplot as plt
plt.bar(regulatory_genes['gene'], regulatory_genes['expression_level'])
plt.title('Niveaux d’expression des gènes régulateurs')
plt.xlabel('Gènes')
plt.ylabel('Niveau d’expression')
plt.show()
```
