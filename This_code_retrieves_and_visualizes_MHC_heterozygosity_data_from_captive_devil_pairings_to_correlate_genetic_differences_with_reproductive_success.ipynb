Load the dataset containing MHC heterozygosity levels and breeding success metrics for captive devil pairs.

```python
import pandas as pd
import plotly.express as px

data = pd.read_csv('devil_mhc_data.csv')
fig = px.scatter(data, x='heterozygosity_difference', y='breeding_success', color='female_age', title='MHC Heterozygosity vs Reproductive Success')
fig.show()
```

The above code plots differences in MHC heterozygosity against reproductive success, enabling visualization of potential disruptive selection effects in the breeding pairs.
