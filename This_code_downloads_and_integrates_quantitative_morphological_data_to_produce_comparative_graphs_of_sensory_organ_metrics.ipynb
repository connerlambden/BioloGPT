This notebook retrieves the provided quantitative data (ampullary pore count and bulb diameter) for Hydrolagus bemisi and Harriotta avia and visualizes it using Plotly.

```python
import plotly.graph_objects as go

species = ['Hydrolagus bemisi', 'Harriotta avia']

ampullary_pore_count = [600, 1245]

ampullary_bulb_diameter = [1434.20, 1167.35]

fig = go.Figure(data=[go.Bar(name='Pore Count', x=species, y=ampullary_pore_count, marker_color='#6A0C76'),
                         go.Bar(name='Bulb Diameter (µm)', x=species, y=ampullary_bulb_diameter, marker_color='#aa66cc')])

fig.update_layout(barmode='group', title='Sensory Organ Comparison', yaxis_title='Counts / Diameter (µm)')

fig.show()
```

The above code generates a grouped bar plot comparing the ampullary pore counts and bulb diameters between the two species, providing visual evidence of the morphological differences discussed.
