This notebook will analyze dosimetric data to compare the effectiveness of IORT applicators and Cerrobend blocks.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('dosimetric_data.csv')
plt.figure(figsize=(10,6))
plt.bar(data['Applicator'], data['Normal_Tissue_Dose'])
plt.title('Normal Tissue Dose Comparison')
plt.xlabel('Applicator Type')
plt.ylabel('Dose (Gy)')
plt.show()
```
