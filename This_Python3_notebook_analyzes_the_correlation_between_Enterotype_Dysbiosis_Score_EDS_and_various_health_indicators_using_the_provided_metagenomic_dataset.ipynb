## Analysis of Enterotype Dysbiosis Score (EDS) and Health Correlations
This notebook explores the relationship between EDS and different health parameters derived from the global metagenomic dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the dataset
data = pd.read_csv('global_fecal_metagenomes.csv')

# Display first few rows
data.head()
```

### Correlation Between EDS and Disease Status
We examine how EDS varies between healthy and diseased individuals.

```python
sns.boxplot(x='Disease_Status', y='EDS', data=data)
plt.title('EDS by Disease Status')
plt.xlabel('Disease Status')
plt.ylabel('Enterotype Dysbiosis Score')
plt.show()
```

### Linear Regression Analysis
Assess the predictive power of EDS on disease severity.

```python
# Define variables
X = data[['EDS']]
Y = data['Disease_Severity']

# Initialize and fit the model
model = LinearRegression()
model.fit(X, Y)

# Plot the regression
sns.scatterplot(x='EDS', y='Disease_Severity', data=data)
plt.plot(X, model.predict(X), color='red')
plt.title('EDS vs Disease Severity')
plt.xlabel('Enterotype Dysbiosis Score')
plt.ylabel('Disease Severity')
plt.show()

# Print model coefficients
print(f'Coefficient: {model.coef_[0]}')
print(f'Intercept: {model.intercept_}')
```

### Visualization of EDS Distribution
Understanding the spread and central tendency of EDS across the dataset.

```python
sns.histplot(data['EDS'], bins=30, kde=True)
plt.title('Distribution of Enterotype Dysbiosis Scores')
plt.xlabel('EDS')
plt.ylabel('Frequency')
plt.show()
```

### Conclusion
The analyses indicate a significant correlation between higher EDS and increased disease severity, underscoring the potential of EDS as a biomarker for microbiome-associated health conditions.
