### Analysis of CDK Expression in Leishmania
This notebook will analyze RNA-seq data to identify the expression levels of cyclin-dependent kinases in Leishmania.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load RNA-seq data
data = pd.read_csv('leishmania_cdk_expression.csv')

# Plot expression levels
plt.figure(figsize=(10,6))
plt.bar(data['CDK'], data['Expression_Level'], color='blue')
plt.title('CDK Expression Levels in Leishmania')
plt.xlabel('CDK')
plt.ylabel('Expression Level')
plt.show()
```

### Discussion
This analysis provides insights into which CDKs are most highly expressed in Leishmania, guiding future therapeutic strategies.

```python
# Save the plot
plt.savefig('cdk_expression_plot.png')
```
