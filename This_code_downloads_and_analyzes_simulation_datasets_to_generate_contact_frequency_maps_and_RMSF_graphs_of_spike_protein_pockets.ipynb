The following notebook cells illustrate the process of downloading the simulation datasets and utilizing Python libraries such as pandas, numpy, and matplotlib to generate detailed RMSF and contact frequency maps for the SARS-CoV-2 spike protein.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Download simulation dataset (assuming API endpoint and dataset URL provided by the paper)
data_url = 'https://datarepository.example.com/sars_cov2_simulation_data.csv'
df = pd.read_csv(data_url)

# Compute RMSF for the RBD region
rbd_data = df[df['domain'] == 'RBD']
rmsf = rbd_data.groupby('residue')['fluctuation'].mean()

# Plot RMSF
plt.figure(figsize=(10,6))
plt.plot(rmsf.index, rmsf.values, label='RBD RMSF', color='#6A0C76')
plt.xlabel('Residue Number')
plt.ylabel('RMSF (Å)')
plt.title('RMSF of RBD in SARS-CoV-2 Spike Protein')
plt.legend()
plt.grid(True)
plt.show()
```
