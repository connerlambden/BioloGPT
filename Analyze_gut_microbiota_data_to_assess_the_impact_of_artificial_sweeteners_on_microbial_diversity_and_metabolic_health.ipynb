This notebook will analyze gut microbiota data from participants consuming artificial sweeteners versus those consuming natural sugars.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load microbiota data
data = pd.read_csv('microbiota_data.csv')

# Analyze diversity indices
sns.boxplot(x='sweetener_type', y='diversity_index', data=data)
plt.title('Microbiota Diversity by Sweetener Type')
plt.show()
```

The analysis will provide insights into how artificial sweeteners affect gut microbiota diversity.

```python
# Further analysis can include correlation with metabolic markers
correlation = data.corr()
sns.heatmap(correlation, annot=True)
plt.title('Correlation Matrix of Microbiota and Metabolic Markers')
plt.show()
```
