Análisis de expresión génica en pancreatitis.

```python
import pandas as pd
# Cargar datos de expresión génica
exp_data = pd.read_csv('data/pancreatitis_expression.csv')
# Análisis de datos
print(exp_data.describe())
```

Este análisis proporciona información sobre la expresión de genes en condiciones de pancreatitis.

```python
# Visualización de datos
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.bar(exp_data['Gene'], exp_data['Expression'])
plt.title('Expresión génica en pancreatitis')
plt.xlabel('Genes')
plt.ylabel('Nivel de expresión')
plt.show()
```
