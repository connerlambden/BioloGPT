### Lipase Activity Analysis
This notebook analyzes the lipase activity data extracted from various studies to identify trends and correlations.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for lipase activity
activity_data = {
    'Time (hours)': ['1', '3', '7', '24'],
    'Carica pentagona': [57.1, 72.3, 73.9, 82.5],
    'Carica papaya': [53.4, 70.2, 76.5, 84.0],
    'Commercially crude papain': [16.5, 28.9, 48.4, 60.2]
}

# Create DataFrame
df = pd.DataFrame(activity_data)

# Plotting
plt.figure(figsize=(10, 6))
for column in df.columns[1:]:
    plt.plot(df['Time (hours)'], df[column], marker='o', label=column)
plt.title('Lipase Activity Over Time')
plt.xlabel('Time (hours)')
plt.ylabel('Activity (%)')
plt.legend()
plt.grid()
plt.show()
```

### Discussion
This analysis provides insights into the performance of different lipases over time, highlighting their potential applications in various biotechnological processes.
