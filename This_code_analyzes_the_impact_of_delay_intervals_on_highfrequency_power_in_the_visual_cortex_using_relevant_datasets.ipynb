### Data Analysis of High-Frequency Power in Visual Cortex

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for analysis
data = {'Frequency': ['10 Hz', '24 Hz', '40 Hz'],
        'Mean_PPC_BW': [0.0879, 0.0251, 0.0012],
        'Mean_PPC_NW': [0.0870, 0.0138, 0.0055]}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Frequency'], df['Mean_PPC_BW'], label='Mean PPC BW')
plt.bar(df['Frequency'], df['Mean_PPC_NW'], label='Mean PPC NW', alpha=0.7)
plt.title('High Frequency Power in Visual Cortex')
plt.xlabel('Frequency')
plt.ylabel('Mean PPC')
plt.legend()
plt.show()
```

This analysis provides insights into how different frequencies of visual stimuli affect high-frequency power in the visual cortex.
