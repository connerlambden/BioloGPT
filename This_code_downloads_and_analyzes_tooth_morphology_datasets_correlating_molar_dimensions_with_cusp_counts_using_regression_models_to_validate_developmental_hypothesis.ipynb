We begin by loading the fossil and in vitro data on arvicoline tooth dimensions and cusp counts.

```python
import pandas as pd
import statsmodels.api as sm

data = pd.read_csv('path/to/vole_tooth_data.csv')
X = data[['molar_length', 'molar_width']]
X = sm.add_constant(X)
y = data['cusp_count']
model = sm.GLM(y, X, family=sm.families.Poisson()).fit()
print(model.summary())
```

The output validates the relationship between molar dimensions and cusp number, supporting the concept of developmental regulation.

```python
# Plotting (using Plotly)
import plotly.express as px
fig = px.scatter(data, x='molar_length', y='cusp_count', color='molar_width', title='Cusp Number vs Molar Length')
fig.show()
```
