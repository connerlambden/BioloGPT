### Analysis of Tensile Strength Data
This notebook analyzes the tensile strength data collected from the electronic stethoscope samples, providing insights into material performance.

```python
import pandas as pd
import plotly.express as px

data = {
    'Test': ['Test 1', 'Test 2', 'Test 3', 'Test 4', 'Test 5', 'Test 6', 'Test 7', 'Test 8', 'Test 9', 'Test 10'],
    'Tensile Strength (MPa)': [36.01, 34.54, 36.25, 35.95, 35.81, 36.59, 35.81, 35.89, 35.78, 36.21]
}

df = pd.DataFrame(data)
fig = px.bar(df, x='Test', y='Tensile Strength (MPa)', title='Tensile Strength of Stethoscope Samples')
fig.show()
```
