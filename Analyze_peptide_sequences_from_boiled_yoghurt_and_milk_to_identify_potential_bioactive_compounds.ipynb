This notebook will analyze peptide sequences derived from the boiled yoghurt and milk mixture to identify bioactive compounds.

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load peptide data
peptide_data = pd.read_csv('peptide_sequences.csv')

# Analyze sequences for bioactivity
bioactive_peptides = peptide_data[peptide_data['activity'] == 'active']

# Display results
bioactive_peptides.head()
```

The analysis will provide insights into the bioactive potential of the peptides identified.

```python
# Further analysis and visualization of bioactive peptides
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(bioactive_peptides['activity_level'], bins=20)
plt.title('Distribution of Bioactive Peptide Activity Levels')
plt.xlabel('Activity Level')
plt.ylabel('Frequency')
plt.show()
```
