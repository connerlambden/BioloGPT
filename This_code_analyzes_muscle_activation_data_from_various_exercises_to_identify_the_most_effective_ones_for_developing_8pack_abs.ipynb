### Muscle Activation Analysis
This notebook will analyze the muscle activation data from different abdominal exercises.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Exercise': ['Suspended Prone Plank', 'Suspended Roll-Out Plank', 'Stable Prone Plank', 'Stable Roll-Out Plank'],
        'Upper Abs': [3, 4, 2, 3],
        'Lower Abs': [2, 4, 1, 2],
        'Obliques': [4, 4, 1, 2]}

df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Exercise'], df['Upper Abs'], label='Upper Abs')
plt.bar(df['Exercise'], df['Lower Abs'], label='Lower Abs', bottom=df['Upper Abs'])
plt.bar(df['Exercise'], df['Obliques'], label='Obliques', bottom=df['Upper Abs'] + df['Lower Abs'])
plt.ylabel('Activation Level')
plt.title('Muscle Activation Levels During Exercises')
plt.legend()
plt.show()
```
