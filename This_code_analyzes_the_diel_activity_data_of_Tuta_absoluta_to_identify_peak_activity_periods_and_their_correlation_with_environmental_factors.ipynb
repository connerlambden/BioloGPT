### Diel Activity Analysis of Tuta absoluta

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for Tuta absoluta activity
activity_data = {'Time Period': ['20:30-21:30', '19:30-20:30', '21:30-22:30'], 'Number of Adults': [50, 30, 40]}
df = pd.DataFrame(activity_data)

# Plotting the activity
plt.bar(df['Time Period'], df['Number of Adults'], color='blue')
plt.title('Diel Activity of Tuta absoluta')
plt.xlabel('Time Period')
plt.ylabel('Number of Adults Caught')
plt.show()
```

This analysis provides insights into the peak activity times of Tuta absoluta, which can inform pest management strategies.
