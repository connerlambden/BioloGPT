Load the dataset detailing chick performance under different test conditions and analyze group differences.

```python
import pandas as pd
import plotly.express as px

data = {'Test': ['Sagittal', 'Fronto-Parallel (4L)', 'Fronto-Parallel (4R)'],
        'Li-chicks': [36.38, 28.85, 11.89],
        'Di-chicks': [28.44, 18.96, 19.79]}

df = pd.DataFrame(data)
fig = px.bar(df, x='Test', y=['Li-chicks', 'Di-chicks'], barmode='group',
             title='Chick Performance Across Test Conditions',
             labels={'value':'Mean Accuracy', 'Test':'Test Condition'})
fig.show()
```
