This notebook will analyze WBC and ESR data from children to identify correlations with infections and nutritional status.

```python
import pandas as pd

# Load relevant datasets
data = pd.read_csv('wbc_esr_data.csv')

# Analyze correlations
correlation = data.corr()
```

The analysis will provide insights into how WBC and ESR values correlate with various health indicators.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize correlations
sns.heatmap(correlation, annot=True)
plt.title('Correlation between WBC, ESR, and Health Indicators')
plt.show()
```
