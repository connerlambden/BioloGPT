This section describes the process of loading experimental and simulation datasets from the provided GitHub repository to analyze tension distributions.

```python
import requests
import pandas as pd
url = 'https://github.com/DianaKhoromskaia/Epitheli-alShell/raw/main/data/simulation_results.csv'
df = pd.read_csv(url)
df.head()
```

The following code snippet processes the dataset to generate phase diagrams correlating gradient strength with deformation magnitude.

```python
import plotly.express as px
fig = px.line(df, x='gradient_strength', y='deformation', title='Phase Diagram of Epithelial Morphogenesis', markers=True)
fig.show()
```

This completes the notebook demonstration for simulating active morphogenesis using real data extracted from the study.

```python
# The final cell ensures reproducibility and interactive visualization in a Jupyter environment.
fig.write_html('phase_diagram.html')
print('Phase diagram generated and saved as phase_diagram.html')
```
