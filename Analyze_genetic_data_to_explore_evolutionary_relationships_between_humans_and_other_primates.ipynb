This notebook will analyze genetic sequences to identify similarities and differences between humans and other primates.

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load genetic data
human_data = pd.read_csv('human_genome.csv')
primate_data = pd.read_csv('primate_genomes.csv')
```

The following analysis will compare genetic markers between humans and selected primates.

```python
# Compare genetic markers
comparison = human_data.merge(primate_data, on='gene_id')
comparison['similarity'] = comparison['human_sequence'] == comparison['primate_sequence']

# Calculate similarity percentage
similarity_percentage = comparison['similarity'].mean() * 100
similarity_percentage
```

This analysis provides insights into the genetic relationships between humans and other primates.

```python
# Display results
print(f'Similarity percentage: {similarity_percentage:.2f}%')
```
