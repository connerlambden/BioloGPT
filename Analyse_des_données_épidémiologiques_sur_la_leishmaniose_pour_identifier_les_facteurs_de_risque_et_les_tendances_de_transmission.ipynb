Analyse des données épidémiologiques sur la leishmaniose.

```python
import pandas as pd
# Charger les données épidémiologiques
data = pd.read_csv('leishmaniasis_data.csv')
# Analyse des tendances de transmission
trends = data.groupby('year').mean()
```

Visualisation des tendances de transmission.

```python
import matplotlib.pyplot as plt
plt.plot(trends.index, trends['cases'])
plt.title('Tendances de transmission de la leishmaniose')
plt.xlabel('Année')
plt.ylabel('Nombre de cas')
plt.show()
```
