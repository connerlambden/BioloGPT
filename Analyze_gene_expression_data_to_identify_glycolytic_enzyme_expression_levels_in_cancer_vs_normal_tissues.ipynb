This notebook will analyze RNA-seq data to compare glycolytic enzyme expression in cancerous and normal tissues.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
cancer_data = pd.read_csv('cancer_expression.csv')
normal_data = pd.read_csv('normal_expression.csv')

# Merge datasets
merged_data = pd.merge(cancer_data, normal_data, on='gene_id')

# Plot expression levels
plt.figure(figsize=(10,6))
sns.boxplot(data=merged_data, x='condition', y='expression', hue='gene_id')
plt.title('Glycolytic Enzyme Expression in Cancer vs Normal Tissues')
plt.show()
```

This analysis will help identify which glycolytic enzymes are upregulated in cancer tissues compared to normal tissues.
