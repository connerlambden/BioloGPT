This notebook downloads detailed photothermal and flow cytometry data from the study and performs regression analysis to correlate electron transfer efficiency with macrophage phenotype transitions.

```python
import pandas as pd
import numpy as np
from scipy.stats import linregress

# Assume data.csv contains columns: concentration, temperature_rise, SOD_activity, M1_ratio, M2_ratio
# Reading data
data = pd.read_csv('data.csv')

# Perform regression between temperature rise and M2 ratio
slope, intercept, r_value, p_value, std_err = linregress(data['temperature_rise'], data['M2_ratio'])
print('Regression r_value:', r_value)

# Plot the data using matplotlib
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(data['temperature_rise'], data['M2_ratio'], color='purple')
plt.plot(data['temperature_rise'], intercept + slope*data['temperature_rise'], 'r', label='fitted line')
plt.xlabel('Temperature Rise (°C)')
plt.ylabel('M2 Macrophage Ratio')
plt.title('Correlation Between Photothermal Effect and M2 Polarization')
plt.legend()
plt.show()
```
