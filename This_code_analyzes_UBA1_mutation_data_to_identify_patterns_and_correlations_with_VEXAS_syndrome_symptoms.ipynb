This notebook will analyze UBA1 mutation data from various studies to identify common mutations associated with VEXAS syndrome.

```python
import pandas as pd

# Load UBA1 mutation data
data = pd.read_csv('uba1_mutations.csv')

# Analyze mutation frequencies
mutation_counts = data['mutation'].value_counts()

# Display results
mutation_counts
```

The analysis reveals the most common UBA1 mutations associated with VEXAS syndrome, providing insights for further research.

```python
# Visualize mutation frequencies
import matplotlib.pyplot as plt

mutation_counts.plot(kind='bar')
plt.title('Frequency of UBA1 Mutations in VEXAS Syndrome')
plt.xlabel('Mutations')
plt.ylabel('Frequency')
plt.show()
```
