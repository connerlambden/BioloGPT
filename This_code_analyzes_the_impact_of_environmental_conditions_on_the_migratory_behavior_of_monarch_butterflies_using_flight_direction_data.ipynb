### Monarch Butterfly Migration Analysis
This notebook analyzes the flight direction data of monarch butterflies under various environmental conditions.

```python
import pandas as pd

# Sample data for analysis
data = {
    'Group': ['Autumn field generation', 'Summer field generation', 'Autumn growth chamber', 'Autumn lab line'],
    'Mean Vector Angle': [240, 134, 214, 200],
    'Sample Size': [20, 14, 23, 21]
}
df = pd.DataFrame(data)

# Calculate average flight direction
average_angle = df['Mean Vector Angle'].mean()
print(f'Average Flight Direction: {average_angle}')
```
