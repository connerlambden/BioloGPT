This notebook will analyze gene expression profiles from HTLV-1 infected T-cells, focusing on splicing events and metabolic pathways.

```python
import pandas as pd
import numpy as np

# Load gene expression data
gene_data = pd.read_csv('htlv1_gene_expression.csv')

# Analyze splicing events
gene_data['splicing_events'] = gene_data['exon_usage'].apply(lambda x: x.count(','))

# Identify metabolic pathways
metabolic_pathways = gene_data[gene_data['pathway'] == 'metabolism']

# Output results
metabolic_pathways.to_csv('metabolic_pathways_analysis.csv')
```

This analysis will help identify key splicing events and metabolic pathways that are altered in HTLV-1 infected T-cells.
