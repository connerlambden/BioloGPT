This notebook begins by importing macrophage polarization data extracted from relevant GVHD literature and public datasets.

```python
import pandas as pd
import plotly.express as px
# Download dataset (placeholder URL)
df = pd.read_csv('https://example.com/gvhd_macrophage_data.csv')
fig = px.line(df, x='Time', y='Cytokine_Level', color='Macrophage_Type', title='Macrophage Cytokine Trends in GVHD')
fig.show()
```

The code above visualizes trends in cytokine expression, stratified by M1 and M2 markers, facilitating a comparative analysis of their temporal profiles in GVHD.

```python
# Additional analysis
summary = df.groupby('Macrophage_Type')['Cytokine_Level'].describe()
print(summary)
```
