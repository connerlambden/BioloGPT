Análisis de datos de expresión génica en raíces para identificar genes asociados con el crecimiento de raíces laterales.

```python
import pandas as pd
# Cargar datos de expresión génica
exp_data = pd.read_csv('data/exp_data.csv')
# Filtrar genes relacionados con raíces laterales
lateral_root_genes = exp_data[exp_data['trait'] == 'lateral_root']
```

Este análisis permite identificar genes que podrían ser manipulados para mejorar el crecimiento de raíces laterales.

```python
# Visualizar resultados
import matplotlib.pyplot as plt
plt.hist(lateral_root_genes['expression_level'])
plt.title('Distribución de niveles de expresión de genes de raíces laterales')
plt.xlabel('Nivel de expresión')
plt.ylabel('Frecuencia')
plt.show()
```
