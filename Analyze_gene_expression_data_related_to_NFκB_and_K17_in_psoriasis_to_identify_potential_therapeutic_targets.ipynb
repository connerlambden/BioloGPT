This notebook will analyze gene expression data from psoriasis patients to explore the relationship between NF-κB and K17.

```python
import pandas as pd

# Load gene expression data
data = pd.read_csv('psoriasis_gene_expression.csv')

# Analyze NF-κB and K17 expression levels
nfkb_expression = data[data['gene'] == 'NF-kB']
k17_expression = data[data['gene'] == 'K17']

# Display results
print(nfkb_expression)
print(k17_expression)
```
