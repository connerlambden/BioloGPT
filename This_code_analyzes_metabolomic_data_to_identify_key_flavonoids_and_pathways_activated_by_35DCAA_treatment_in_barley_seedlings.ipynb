### Metabolomic Data Analysis for Barley Seedlings

```python
import pandas as pd

# Load metabolomic data
data = pd.read_csv('metabolomic_data.csv')

# Analyze flavonoid accumulation
flavonoids = data[data['compound'].str.contains('flavonoid')]
flavonoid_summary = flavonoids.groupby('time_point').sum()
flavonoid_summary
```

### Summary of Flavonoid Accumulation Over Time

```python
import matplotlib.pyplot as plt

# Plotting flavonoid accumulation
times = flavonoid_summary.index
values = flavonoid_summary['amount']

plt.plot(times, values)
plt.title('Flavonoid Accumulation Over Time')
plt.xlabel('Time (h)')
plt.ylabel('Flavonoid Amount')
plt.show()
```
