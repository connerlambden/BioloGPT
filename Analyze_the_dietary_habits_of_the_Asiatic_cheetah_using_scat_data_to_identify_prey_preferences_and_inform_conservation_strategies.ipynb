This notebook analyzes scat data to determine the dietary habits of the Asiatic cheetah.

```python
import pandas as pd

diet_data = {
    'Food Item': ['Goitered Gazelle', 'Wild Boar', 'Livestock', 'Hare', 'Rodents', 'Reptiles', 'Insects', 'Plant Material'],
    'Miandasht Frequency (%)': [26, 0, 57, 6, 34, 6, 9, 40],
    'Behkadeh Frequency (%)': [39, 22, 39, 13, 35, 0, 4, 39]
}

diet_df = pd.DataFrame(diet_data)
diet_df.set_index('Food Item', inplace=True)
diet_df
```

This analysis provides insights into the dietary preferences of the Asiatic cheetah, which can guide conservation efforts.

```python
diet_df.plot(kind='bar', title='Dietary Preferences of Asiatic Cheetah', ylabel='Frequency of Occurrence (%)')
```
