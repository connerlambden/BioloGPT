### Data Analysis of Stickies in Recycling
This notebook analyzes the effects of temperature and pulp consistency on the size and removal rates of stickies.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Temperature (°C)': [70, 80, 90, 100],
        'Removal (%)': [20.1, 21.3, 24.8, 25.7],
        'Size (mm²)': [0.04, 0.038, 0.034, 0.028]}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 5))
plt.bar(df['Temperature (°C)'], df['Removal (%)'], color='blue', alpha=0.6, label='Removal (%)')
plt.plot(df['Temperature (°C)'], df['Size (mm²)'], color='red', marker='o', label='Size (mm²)')
plt.title('Effect of Temperature on Stickies Size and Removal')
plt.xlabel('Temperature (°C)')
plt.ylabel('Size (mm²) / Removal (%)')
plt.legend()
plt.show()
```
