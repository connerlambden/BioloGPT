## Transcriptome Analysis of Dual-Species Biofilms

This notebook analyzes RNA-seq data from Candida auris co-cultured with different Candida species to identify genes involved in biofilm interactions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
```

### Loading the Data

We will load the transcriptome data for C. auris when co-cultured with C. albicans and C. krusei.

```python
data_albicans = pd.read_csv('E-MTAB-12667/data_albicans.csv')
data_krusei = pd.read_csv('E-MTAB-12667/data_krusei.csv')
```

### Differential Gene Expression Analysis

Identify genes significantly upregulated or downregulated in co-culture conditions.

```python
threshold = 0.05
genes = data_albicans['gene'].append(data_krusei['gene']).unique()
diff_expr = []
for gene in genes:
    expr_alb = data_albicans[data_albicans['gene'] == gene]['expression']
    expr_krus = data_krusei[data_krusei['gene'] == gene]['expression']
    stat, p = ttest_ind(expr_alb, expr_krus)
    if p < threshold:
        diff_expr.append((gene, p))
diff_expr_df = pd.DataFrame(diff_expr, columns=['Gene', 'p-value'])
print(diff_expr_df.head())
```

### Visualization of Differential Expression

Plot the top differentially expressed genes.

```python
top_genes = diff_expr_df.nsmallest(10, 'p-value')
sns.barplot(x='Gene', y='p-value', data=top_genes)
plt.xticks(rotation=45)
plt.title('Top Differentially Expressed Genes in C. auris Co-culture')
plt.show()
```

### Conclusion

The analysis highlights key genes that may drive the competitive interactions observed in dual-species biofilms.

```python
# Save the differential expression results for further analysis
diff_expr_df.to_csv('differential_expression_results.csv', index=False)
```
