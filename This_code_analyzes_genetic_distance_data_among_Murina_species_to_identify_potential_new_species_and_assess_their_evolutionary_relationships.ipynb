### Genetic Distance Analysis of Murina Species
This notebook analyzes genetic distance data to identify potential new species within the Murina genus.

```python
import pandas as pd
import matplotlib.pyplot as plt

distance_data = {'Species': ['M. lvchun', 'M. pluvialis', 'M. cyclotis'], 'Genetic Distance': [7.3, 0, 12.5]}
df = pd.DataFrame(distance_data)

plt.bar(df['Species'], df['Genetic Distance'])
plt.title('Genetic Distances Among Murina Species')
plt.xlabel('Species')
plt.ylabel('Genetic Distance (%)')
plt.show()
```

### Discussion
This analysis highlights the significant genetic distances among Murina species, supporting the identification of new species.
