### Data Analysis of Voltage and Impedance in Atrial Fibrillation Patients

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {'Voltage': [0.5, 0.4, 0.3, 0.2, 0.1], 'Impedance': [151.5, 129.1, 140.0, 135.0, 145.0]}
df = pd.DataFrame(data)

# Correlation analysis
correlation = df.corr()

# Visualization
sns.scatterplot(x='Voltage', y='Impedance', data=df)
plt.title('Voltage vs Impedance in AF Patients')
plt.show()
```
