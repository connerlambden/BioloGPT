### Step 1: Import Libraries
This section imports necessary libraries for data handling and deep learning.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils import to_categorical
```

### Step 2: Load and Preprocess Data
Load the MALDI-ToF dataset and preprocess it for training.

```python
# Load dataset
data = pd.read_csv('maldi_data.csv')
X = data.drop('label', axis=1).values  # Features
y = data['label'].values  # Labels

# Encode labels
y_encoded = to_categorical(y)

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```
