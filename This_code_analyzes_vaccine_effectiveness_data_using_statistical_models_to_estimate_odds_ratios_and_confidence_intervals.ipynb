### Vaccine Effectiveness Analysis
This notebook will guide you through the analysis of vaccine effectiveness data using logistic regression.

```python
import pandas as pd
import statsmodels.api as sm

# Load your dataset
data = pd.read_csv('vaccine_data.csv')

# Define the logistic regression model
model = sm.Logit(data['outcome'], data[['vaccinated', 'age', 'health_status']])
result = model.fit()

# Display the results
print(result.summary())
```

### Results Interpretation
The output will provide the odds ratios for the vaccinated group compared to the unvaccinated group, allowing for the calculation of vaccine effectiveness.

```python
# Calculate vaccine effectiveness
odds_ratio = result.params['vaccinated']
ve = 100 * (1 - odds_ratio)
print(f'Vaccine Effectiveness: {ve}%')
```
