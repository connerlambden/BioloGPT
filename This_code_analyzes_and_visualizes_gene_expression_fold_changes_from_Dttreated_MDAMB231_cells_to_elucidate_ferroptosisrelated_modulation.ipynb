Import necessary libraries and load the dataset containing gene expression fold change data for GPX4, TFRC, and SLC7A11 at 1, 3, and 24 hours.

```python
import pandas as pd
import plotly.express as px

data = {'Time': ['1 h', '3 h', '24 h'],
        'GPX4': [-28.72, -20.17, -31.1],
        'TFRC': [72.00, 155.00, 227.43],
        'SLC7A11': [-12.57, -4.48, -13.38]}

df = pd.DataFrame(data)

fig = px.bar(df, x='Time', y=['GPX4', 'TFRC', 'SLC7A11'], barmode='group', title='Gene Expression Fold Change')
fig.show()
```
