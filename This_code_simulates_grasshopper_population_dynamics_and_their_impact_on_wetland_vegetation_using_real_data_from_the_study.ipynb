### Grasshopper Population Dynamics Simulation
This notebook simulates the impact of grasshopper populations on wetland vegetation based on the findings from the study.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
initial_biomass = 1000  # Initial biomass of wetland
grasshopper_density = np.arange(1, 31, 1)  # Density from 1 to 30 individuals/m²
biomass_loss = 50 * (grasshopper_density / 30)  # Linear loss model

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(grasshopper_density, biomass_loss, marker='o')
plt.title('Biomass Loss vs Grasshopper Density')
plt.xlabel('Grasshopper Density (individuals/m²)')
plt.ylabel('Biomass Loss (%)')
plt.grid()
plt.show()
```
