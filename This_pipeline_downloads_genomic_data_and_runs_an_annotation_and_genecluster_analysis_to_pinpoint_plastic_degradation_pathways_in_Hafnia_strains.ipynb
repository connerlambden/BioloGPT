Download genome sequences using provided BioProject IDs and run Prokka for gene annotation.

```python
import subprocess
# Example command to annotate a genome
subprocess.call(['prokka', '--outdir', 'annotation', '--prefix', 'hafnia', 'genome.fasta'])
print('Annotation complete.')
```

Load annotated GFF file and filter for genes related to plastic degradation using keyword searches.

```python
import pandas as pd
gff_data = pd.read_csv('annotation/hafnia.gff', sep='\t', comment='#', header=None)
degradation_genes = gff_data[gff_data[8].str.contains('degrad', na=False)]
print(degradation_genes.head())
```
