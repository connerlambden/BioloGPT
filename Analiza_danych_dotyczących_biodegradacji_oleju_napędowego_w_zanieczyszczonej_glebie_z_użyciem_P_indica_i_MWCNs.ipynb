Zbieranie danych dotyczących biodegradacji oleju napędowego w glebie zanieczyszczonej Pb i Cd.

```python
import pandas as pd

# Wczytanie danych
data = pd.read_csv('biodegradation_data.csv')

# Analiza danych
summary = data.describe()
```

Analiza statystyczna wyników biodegradacji w różnych warunkach.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Wizualizacja danych
sns.barplot(x='treatment', y='biodegradation_increase', data=data)
plt.title('Wzrost biodegradacji w różnych warunkach')
plt.show()
```

```python
# Podsumowanie wyników
print(summary)
```
