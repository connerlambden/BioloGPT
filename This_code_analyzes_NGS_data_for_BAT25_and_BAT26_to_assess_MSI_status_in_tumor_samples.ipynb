### Step 1: Import Required Libraries
Import necessary libraries for data analysis and visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load Data
Load the NGS data containing BAT-25 and BAT-26 results.

```python
data = pd.read_csv('ngs_data.csv')
print(data.head())
```

### Step 3: Analyze MSI Status
Analyze the MSI status based on BAT-25 and BAT-26 results.

```python
msi_status = data[['BAT-25', 'BAT-26']].apply(lambda x: 'MSI-H' if (x['BAT-25'] == 'unstable' or x['BAT-26'] == 'unstable') else 'MSS', axis=1)
data['MSI_Status'] = msi_status
print(data[['Sample_ID', 'MSI_Status']])
```

### Step 4: Visualize Results
Visualize the distribution of MSI status in the dataset.

```python
sns.countplot(x='MSI_Status', data=data)
plt.title('Distribution of MSI Status')
plt.xlabel('MSI Status')
plt.ylabel('Count')
plt.show()
```
