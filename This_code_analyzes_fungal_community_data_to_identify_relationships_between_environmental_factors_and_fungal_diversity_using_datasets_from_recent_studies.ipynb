### Fungal Community Analysis
This notebook analyzes the diversity of fungal communities based on environmental factors such as salinity and soil nutrients.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load fungal community data
data = pd.read_csv('fungal_community_data.csv')

# Analyze diversity based on salinity
salinity_diversity = data.groupby('salinity').agg({'ASV_count': 'sum', 'relative_abundance': 'mean'})

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(salinity_diversity.index, salinity_diversity['relative_abundance'], color='skyblue')
plt.title('Fungal Diversity by Salinity')
plt.xlabel('Salinity Level')
plt.ylabel('Average Relative Abundance (%)')
plt.show()
```

### Discussion
This analysis provides insights into how salinity affects fungal diversity, which is crucial for understanding ecosystem dynamics.

```python
# Save the plot
plt.savefig('fungal_diversity_by_salinity.png')
```
