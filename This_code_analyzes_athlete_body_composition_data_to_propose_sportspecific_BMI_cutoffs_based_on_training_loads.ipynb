### Step 1: Data Collection
Collect body composition data from athletes across various sports, focusing on BMI, body fat percentage, and training loads.

```python
import pandas as pd

# Load athlete data
data = pd.read_csv('athlete_body_composition.csv')

# Display the first few rows of the dataset
print(data.head())
```

### Step 2: Data Analysis
Analyze the relationship between training load and body composition metrics.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize the relationship between training load and body fat percentage
sns.scatterplot(data=data, x='training_load', y='body_fat_percentage')
plt.title('Training Load vs Body Fat Percentage')
plt.show()
```

### Step 3: Propose New Cut-Offs
Based on the analysis, propose new BMI cut-offs for different sports.

```python
# Calculate new BMI cut-offs based on body fat percentage
new_cutoffs = data.groupby('sport')['BMI'].agg(['mean', 'std']).reset_index()
new_cutoffs['new_cutoff_overweight'] = new_cutoffs['mean'] + new_cutoffs['std']
new_cutoffs['new_cutoff_obesity'] = new_cutoffs['mean'] + 2 * new_cutoffs['std']

print(new_cutoffs)
```
