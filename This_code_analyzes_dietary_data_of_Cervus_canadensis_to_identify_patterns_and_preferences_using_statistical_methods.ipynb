### Dietary Analysis of Cervus canadensis

```python
import pandas as pd
import matplotlib.pyplot as plt

diet_data = {'Year': [2021, 2022], 'Salix spp.': [46.53, 36.70], 'Rhododendron fragariiflorum': [17.79, 10.11]}
df = pd.DataFrame(diet_data)
df.set_index('Year').plot(kind='bar')
plt.title('Diet Composition of Tibetan Red Deer')
plt.ylabel('Percentage')
plt.show()
```

This analysis visualizes the dietary composition of Tibetan red deer over two years, highlighting changes in food preferences.
