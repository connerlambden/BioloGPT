### Binding Affinity Analysis of Aptamers to RAD51
This analysis will utilize the binding affinity data from the study to visualize the interaction strengths of different aptamers.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for aptamer binding affinities
aptamer_data = {'Aptamer': ['Apt1', 'Apt2', 'Apt3', 'Apt4', 'rcApt'], 'Kd (nM)': [116, 138, 188, 413, 0]}
df = pd.DataFrame(aptamer_data)

# Plotting the binding affinities
plt.figure(figsize=(10,6))
plt.bar(df['Aptamer'], df['Kd (nM)'], color='blue')
plt.title('Binding Affinities of Aptamers to RAD51')
plt.xlabel('Aptamer')
plt.ylabel('Kd (nM)')
plt.grid(axis='y')
plt.show()
```

This plot illustrates the varying binding affinities of the designed aptamers to RAD51, highlighting Apt1 as the most effective candidate.
