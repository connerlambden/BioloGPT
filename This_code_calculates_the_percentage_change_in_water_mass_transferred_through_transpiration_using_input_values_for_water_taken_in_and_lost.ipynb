# Water Mass Calculation
This notebook calculates the percentage change in water mass transferred from roots to atmosphere through transpiration.

```python
# Input values
water_taken_in = 5600  # kg
water_lost = 4364  # kg

# Calculate percentage change
percentage_change = ((water_lost - water_taken_in) / water_taken_in) * 100
percentage_change
```
