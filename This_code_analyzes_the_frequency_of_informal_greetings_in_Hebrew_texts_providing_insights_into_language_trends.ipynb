Analyze Hebrew text data for informal greeting usage.

```python
import pandas as pd
# Load Hebrew text data
data = pd.read_csv('hebrew_texts.csv')
# Count occurrences of 'היי'
hi_count = data['text'].str.count('היי').sum()
```

This analysis helps understand the prevalence of informal greetings in contemporary Hebrew.

```python
print(f'Total occurrences of היי: {hi_count}')
```
