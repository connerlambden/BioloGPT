### RNA-Seq Data Analysis of Myometrial deTARs in Mares with Endometrosis

This notebook processes RNA-Seq data to identify differentially expressed genes and visualize enriched biological pathways related to endometrosis stages.

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import plotly.express as px
from bioinfokit.analys import gprofiler
```

### Load and Prepare Data

Load the RNA-Seq data and prepare it for analysis.

```python
# Load differentially expressed transcriptionally active regions (deTARs)
deTARs_IIA_I = pd.read_csv('deTARs_IIA_vs_I.csv')
deTARs_IIB_I = pd.read_csv('deTARs_IIB_vs_I.csv')
deTARs_III_I = pd.read_csv('deTARs_III_vs_I.csv')
```

### Identify Common deTARs Across All Stages

Find the intersection of deTARs across all comparisons to identify common regulatory changes.

```python
# Identify common deTARs
common_deTARs = set(deTARs_IIA_I['gene']).intersection(set(deTARs_IIB_I['gene']), set(deTARs_III_I['gene']))
common_deTARs_df = pd.DataFrame(list(common_deTARs), columns=['gene'])
common_deTARs_df.head()
```

### Pathway Enrichment Analysis

Perform pathway enrichment to understand the biological processes impacted by the common deTARs.

```python
# Perform pathway enrichment using g:Profiler
gp = gprofiler.GProfiler(return_dataframe=True)
enriched_paths = gp.profile(organism='hsapiens', query=common_deTARs, sources=['GO:BP', 'KEGG'])
enriched_paths.head()
```

### Visualize Enriched Pathways

Create a bar plot to visualize the top enriched pathways.

```python
# Plot top enriched pathways
top_paths = enriched_paths.head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='source', y='name', data=top_paths, palette='viridis')
plt.title('Top Enriched Pathways in Common deTARs')
plt.xlabel('Source')
plt.ylabel('Pathway')
plt.tight_layout()
plt.show()
```
