```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Load proteomic dataset
proteomics_data = pd.read_csv('PXD054348_proteomics.csv')
# Filter data for conditions
control = proteomics_data[proteomics_data['condition'] == 'control']
stimulated = proteomics_data[proteomics_data['condition'] == 'stimulated']

# Statistical analysis
diff_expr = ttest_ind(control['expression'], stimulated['expression'])

plt.figure(figsize=(8,6))
plt.hist(control['expression'], alpha=0.5, label='Control', color='#6A0C76')
plt.hist(stimulated['expression'], alpha=0.5, label='Stimulated', color='#FF5733')
plt.legend()
plt.title('Protein Expression Distribution')
plt.xlabel('Expression Level')
plt.ylabel('Frequency')
plt.show()
print('t-test p-value:', diff_expr.pvalue)
```
