### Wound Healing Data Analysis
This notebook will analyze the wound healing data obtained from the CSMA plugin.

```python
import pandas as pd
import plotly.express as px

# Sample data based on CSMA results
data = {
    'Treatment': ['DMSO', '10 nM SVA', '100 µM WVA'],
    'Mean Linear Wound Closure Rate': [1.63, 1.17, 0.78],
    'Final Wound Area': [59.89, 45.20, 27.22]
}
df = pd.DataFrame(data)

# Create a bar plot for mean linear wound closure rate
fig = px.bar(df, x='Treatment', y='Mean Linear Wound Closure Rate',
             title='Mean Linear Wound Closure Rate by Treatment',
             labels={'Mean Linear Wound Closure Rate': '% per hour'})
fig.show()
```

The analysis provides insights into the effectiveness of different treatments on wound healing.
