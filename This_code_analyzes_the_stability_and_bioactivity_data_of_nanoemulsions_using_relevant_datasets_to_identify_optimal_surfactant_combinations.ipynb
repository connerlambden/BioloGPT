This notebook will analyze the stability and bioactivity of nano-emulsions derived from Curcuma longa essential oils.

```python
import pandas as pd

# Load relevant datasets
dataset = pd.read_csv('nanoemulsion_data.csv')

# Analyze stability and bioactivity
results = dataset.groupby('surfactant_combination').agg({'stability': 'mean', 'bioactivity': 'mean'})
results.to_csv('optimized_results.csv')
```

The results will provide insights into the best surfactant combinations for enhancing the stability and bioactivity of nano-emulsions.
