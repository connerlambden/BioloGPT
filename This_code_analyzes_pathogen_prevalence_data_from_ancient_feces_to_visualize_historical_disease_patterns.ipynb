### Pathogen Prevalence Analysis in Ancient Feces

```python
import pandas as pd
import plotly.express as px

# Data for pathogen prevalence
data = {'Pathogen': ['Blastocystis spp.', 'Atypical E. coli', 'Enterobius vermicularis', 'Entamoeba spp.', 'Enterotoxigenic E. coli', 'Shigella spp.', 'Giardia spp.', 'E. coli O157:H7'],
        'Prevalence (%)': [70, 70, 60, 50, 50, 30, 20, 10]}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
fig = px.bar(df, x='Pathogen', y='Prevalence (%)', title='Pathogen Prevalence in Ancient Feces', color='Prevalence (%)', text='Prevalence (%)')
fig.show()
```
