This notebook will analyze gene expression data from bladder sensory neurons to identify key genes involved in sensory signaling.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('bladder_sensory_genes.csv')
sns.boxplot(x='gene', y='expression', data=data)
plt.title('Gene Expression in Bladder Sensory Neurons')
plt.show()
```
