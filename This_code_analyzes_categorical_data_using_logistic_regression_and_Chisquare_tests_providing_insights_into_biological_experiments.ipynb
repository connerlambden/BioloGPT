### Step 1: Import Libraries
Import necessary libraries for data analysis.

```python
import pandas as pd
import statsmodels.api as sm
from scipy import stats
```

### Step 2: Load Data
Load your categorical dataset for analysis.

```python
data = pd.read_csv('your_data.csv')
```

### Step 3: Logistic Regression Analysis
Perform logistic regression on the dataset.

```python
logit_model = sm.Logit(data['outcome'], data[['predictor1', 'predictor2']])
result = logit_model.fit()
print(result.summary())
```

### Step 4: Chi-Square Test
Conduct a Chi-square test for independence.

```python
contingency_table = pd.crosstab(data['variable1'], data['variable2'])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
print(f'Chi2: {chi2}, p-value: {p}')
```
