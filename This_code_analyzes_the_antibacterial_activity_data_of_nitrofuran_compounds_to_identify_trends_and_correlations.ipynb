### Antibacterial Activity Analysis
This notebook analyzes the MIC data of various nitrofuran compounds against different bacterial strains.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for analysis
data = {'Compound': ['Lead Compound', 'Compound 2b', 'Compound 5c'],
        'MIC_Mtb': [0.0124, 100, 6.2],
        'MIC_S_aureus': [100, 0.171, 0.19]}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 5))
plt.bar(df['Compound'], df['MIC_Mtb'], label='MIC against MTb (µg/mL)', alpha=0.6)
plt.bar(df['Compound'], df['MIC_S_aureus'], label='MIC against S. aureus (µg/mL)', alpha=0.6)
plt.title('Antibacterial Activity of Compounds')
plt.xlabel('Compounds')
plt.ylabel('MIC (µg/mL)')
plt.legend()
plt.show()
```
