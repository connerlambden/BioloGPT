### Data Analysis of Plant Lifespan
This analysis will focus on comparing the genetic diversity of plants grown from seeds and cuttings.

```python
import pandas as pd

# Load relevant datasets
dataset_seeds = pd.read_csv('seeds_data.csv')
dataset_cuttings = pd.read_csv('cuttings_data.csv')

# Analyze genetic diversity
seed_diversity = dataset_seeds['genetic_diversity'].mean()
cutting_diversity = dataset_cuttings['genetic_diversity'].mean()

print(f'Seed Genetic Diversity: {seed_diversity}')
print(f'Cutting Genetic Diversity: {cutting_diversity}')
```
