This notebook details the analysis of RNA-seq data and MIC values to establish a correlation between oxidative stress marker expression and berberrubine efficacy.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load MIC and gene expression datasets
mic_data = pd.DataFrame({'Compound': ['Berberrubine'], 'MIC_Hp_mg_ml': [0.011], 'MIC_Ec_mg_ml': [0.17]})
gene_exp = pd.DataFrame({'Gene': ['fumC', 'sodB'], 'Expression_Change': ['Upregulated', 'Upregulated']})

# Display datasets
print(mic_data)
print(gene_exp)

# Plot MIC values
plt.bar(mic_data['Compound'], mic_data['MIC_Hp_mg_ml'], color='purple', label='H pylori')
plt.bar(mic_data['Compound'], mic_data['MIC_Ec_mg_ml'], color='orange', label='E coli', alpha=0.7)
plt.ylabel('MIC (mg/ml)')
plt.title('MIC Values of Berberrubine')
plt.legend()
plt.show()
```
