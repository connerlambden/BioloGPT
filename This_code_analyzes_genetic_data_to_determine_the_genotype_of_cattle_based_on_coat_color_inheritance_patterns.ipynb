### Genetic Analysis of Cattle Coat Color
This notebook will analyze the inheritance patterns of coat color in cattle using genetic data.

```python
import pandas as pd

def analyze_genotype(data):
    # Analyze the genotype data
    results = data.groupby('Genotype').size()
    return results

# Example data
cattle_data = pd.DataFrame({'Genotype': ['BB', 'Bb', 'Bb', 'bb', 'BB', 'Bb']})

# Analyze the genotype
analyze_genotype(cattle_data)
```
