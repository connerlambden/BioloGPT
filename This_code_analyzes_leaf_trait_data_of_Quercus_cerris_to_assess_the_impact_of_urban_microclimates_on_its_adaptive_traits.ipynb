### Data Analysis of Quercus cerris Leaf Traits
This notebook will analyze leaf trait data to understand how urban microclimates affect Quercus cerris.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
leaf_data = pd.read_csv('quercus_cerris_leaf_traits.csv')

# Display the first few rows of the dataset
leaf_data.head()
```

### Statistical Analysis
We will perform statistical analyses to identify significant differences in leaf traits between urban and natural populations.

```python
from scipy import stats

# Perform t-test between urban and natural leaf traits
urban_traits = leaf_data[leaf_data['Location'] == 'Urban']
natural_traits = leaf_data[leaf_data['Location'] == 'Natural']
t_stat, p_value = stats.ttest_ind(urban_traits['SLA'], natural_traits['SLA'])

print(f'T-statistic: {t_stat}, P-value: {p_value}')
```
