Load and merge datasets from FluNet, WAHIS, and the Influenza Research Database. Analyze key metrics (incidence, DALYs, testing rates) per country.

```python
import pandas as pd
# Code to download and merge datasets
flu_data = pd.read_csv('flu_data.csv')
animal_data = pd.read_csv('animal_data.csv')
merged_data = pd.merge(flu_data, animal_data, on='country', how='outer')
print(merged_data.head())
```

Visualize the disparities with a Plotly bar chart comparing incidence rates and DALYs.

```python
import plotly.express as px
fig = px.bar(merged_data, x='country', y=['incidence_rate', 'DALYs'], barmode='group', title='Incidence vs DALYs in LAC')
fig.show()
```
