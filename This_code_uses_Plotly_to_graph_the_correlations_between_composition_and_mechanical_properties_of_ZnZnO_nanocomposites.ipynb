Step-by-step analysis of Zn-ZnO composite data, mapping composition to microhardness and tensile strength.

```python
import plotly.graph_objs as go
import plotly.offline as pyo

# Define composition categories and properties
categories = ['Pure Zn', 'Zn-1% ZnO', 'Zn-3% ZnO', 'Zn-5% ZnO']
microhardness = [42.3, 46.0, 60.0, 80.0]
tensile_strength = [32.1, 35.6, 60.6, 84.3]

# Create bar traces
data = [
    go.Bar(x=categories, y=microhardness, name='Microhardness (HV)', marker_color='#6A0C76'),
    go.Bar(x=categories, y=tensile_strength, name='Tensile Strength (MPa)', marker_color='lightslategray')
]

# Define layout
layout = go.Layout(title='Mechanical Properties of Zn-ZnO Nanocomposites', barmode='group')

# Plot the graph
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig, filename='composite_properties.html')
```
