Este análisis utilizará datos de expresión génica y metabolómica para evaluar cómo diferentes organismos obtienen y utilizan energía.

```python
import pandas as pd
import numpy as np

# Cargar datos de expresión génica y metabolómica
gene_expression = pd.read_csv('gene_expression_data.csv')
metabolomics = pd.read_csv('metabolomics_data.csv')

# Análisis de correlación entre expresión génica y producción de energía
correlation = gene_expression.corrwith(metabolomics['energy_production'])
print(correlation)
```
