Análisis de datos de clasificación de sustancias químicas.

```python
import pandas as pd

# Cargar datos de clasificación
data = pd.read_csv('hazard_data.csv')

# Análisis de peligrosidad
hazard_analysis = data.groupby('Polímero')['Puntaje de Peligro'].mean()
```

Este análisis permite identificar los polímeros más peligrosos y sus monómeros asociados.

```python
print(hazard_analysis)
```
