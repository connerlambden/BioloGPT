This notebook downloads the imaging performance data and uses Plotly to visualize spatial resolution against imaging depth.

```python
import plotly.graph_objects as go
import pandas as pd

# Define the data
data = {'Imaging Depth': ['230 μm', '415 μm', '615 μm'],
        'Lateral Resolution (μm)': [1.24, 1.31, 1.44],
        'Axial Resolution (μm)': [16.78, 19.62, 26.89]}
df = pd.DataFrame(data)

# Create bar plots
fig = go.Figure(data=[
    go.Bar(name='Lateral Resolution', x=df['Imaging Depth'], y=df['Lateral Resolution (μm)'], marker_color='#A020F0'),
    go.Bar(name='Axial Resolution', x=df['Imaging Depth'], y=df['Axial Resolution (μm)'], marker_color='#6A0C76')
])

# Change the bar mode
fig.update_layout(barmode='group', title='Spatial Resolution vs Imaging Depth', xaxis_title='Imaging Depth', yaxis_title='Resolution (μm)')
fig.show()
```

The above code visualizes the trade-off between lateral and axial resolution across different imaging depths.
