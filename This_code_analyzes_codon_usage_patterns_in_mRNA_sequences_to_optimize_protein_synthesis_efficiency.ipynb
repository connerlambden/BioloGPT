### Codon Usage Analysis
This notebook analyzes codon usage in various mRNA sequences to identify patterns that can enhance protein synthesis efficiency.

```python
import pandas as pd

# Load mRNA sequences
data = pd.read_csv('mRNA_sequences.csv')

# Analyze codon usage
codon_usage = data['sequence'].apply(lambda x: {codon: x.count(codon) for codon in set(x)})

# Display results
codon_usage_summary = pd.DataFrame(codon_usage.tolist()).fillna(0)
print(codon_usage_summary)
```
