### Analyzing Salmonella Antibody Titer Data
This notebook will analyze the relationship between Salmonella antibody titers and patient demographics.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('salmonella_titer_data.csv')

# Analyze the distribution of titers
plt.hist(data['titer'], bins=20)
plt.title('Distribution of Salmonella Antibody Titers')
plt.xlabel('Titer')
plt.ylabel('Frequency')
plt.show()
```
