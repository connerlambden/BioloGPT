### Mutual Fund Performance Analysis
This notebook will analyze historical performance data of various mutual funds to identify those with the best returns and lowest risk.

```python
import pandas as pd

# Load mutual fund performance data
data = pd.read_csv('mutual_fund_data.csv')

# Analyze performance metrics
performance = data[['Fund Name', '1-Year Return', '3-Year Return', 'Expense Ratio']]
performance['Risk'] = performance['1-Year Return'] / performance['Expense Ratio']

# Identify top performers
top_performers = performance.sort_values(by='Risk', ascending=False).head(10)
top_performers
```

### Summary of Top Performing Funds
The analysis will provide insights into which funds have historically provided the best returns relative to their risk.

```python
# Display top performers
print(top_performers)
```
