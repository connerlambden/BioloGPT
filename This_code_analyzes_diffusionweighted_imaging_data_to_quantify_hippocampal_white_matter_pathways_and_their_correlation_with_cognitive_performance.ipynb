### Data Preparation
Load the diffusion-weighted imaging data and cognitive performance metrics.

```python
import pandas as pd
import numpy as np

dwi_data = pd.read_csv('dwi_data.csv')
cognitive_scores = pd.read_csv('cognitive_scores.csv')
```

### Data Analysis
Perform statistical analysis to correlate white matter pathway variability with cognitive performance.

```python
from scipy.stats import pearsonr

results = []
for pathway in dwi_data.columns[1:]:
    correlation, p_value = pearsonr(dwi_data[pathway], cognitive_scores['performance'])
    results.append((pathway, correlation, p_value))

results_df = pd.DataFrame(results, columns=['Pathway', 'Correlation', 'P-Value'])
```

### Results
Display the results of the correlation analysis.

```python
results_df.sort_values(by='Correlation', ascending=False).head(10)
```
