This notebook analyzes gene expression data from heart studies to identify key genes involved in heart development and disease.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('heart_gene_expression.csv')
plt.figure(figsize=(10,6))
plt.bar(data['Gene'], data['Expression'])
plt.title('Gene Expression in Heart Development')
plt.xlabel('Genes')
plt.ylabel('Expression Level')
plt.xticks(rotation=45)
plt.show()
```

The bar chart above illustrates the expression levels of various genes involved in heart development, highlighting potential targets for further research.
