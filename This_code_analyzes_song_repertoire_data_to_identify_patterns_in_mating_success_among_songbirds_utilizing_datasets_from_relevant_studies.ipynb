### Analyzing Song Repertoire Data
This notebook will analyze the relationship between song repertoire size and mating success in songbirds.

```python
import pandas as pd

# Load data
repertoire_data = pd.DataFrame({
    'male_id': [1, 2, 3],
    'repertoire_size': [8, 6, 10],
    'mating_success': ['yes', 'no', 'yes']
})

# Analyze data
success_rate = repertoire_data['mating_success'].value_counts(normalize=True)
print(success_rate) # Display success rates
```

This analysis reveals the proportion of mating success among males with different song repertoire sizes.
