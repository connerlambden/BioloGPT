### Gene Expression Analysis of CEBPB and Helios in Megakaryocytes

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('megakaryocyte_expression_data.csv')
sns.boxplot(x='Condition', y='Expression', data=data)
plt.title('Expression of CEBPB and Helios in Megakaryocytes')
plt.show()
```
