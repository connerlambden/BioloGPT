### Data Preparation
Load MRI data and preprocess it for analysis.

```python
import pandas as pd
import numpy as np

# Load MRI data
mri_data = pd.read_csv('mri_data.csv')
# Preprocess data
mri_data_cleaned = mri_data.dropna()
```

### Network Analysis
Construct brain networks and calculate properties.

```python
import networkx as nx

# Create a graph from the cleaned MRI data
G = nx.from_pandas_adjacency(mri_data_cleaned.corr())

# Calculate network properties
small_worldness = nx.small_worldness(G)
clustering_coefficient = nx.average_clustering(G)
```
