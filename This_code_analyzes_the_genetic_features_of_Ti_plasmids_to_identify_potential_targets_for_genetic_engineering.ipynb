### Analysis of Ti Plasmid Features

```python
# Import necessary libraries
import pandas as pd

# Load Ti plasmid data
plasmid_data = pd.read_csv('ti_plasmid_data.csv')

# Display the data
plasmid_data.head()
```

### Summary Statistics

```python
# Generate summary statistics
summary_stats = plasmid_data.describe()
summary_stats
```

### Visualization of Plasmid Features

```python
import matplotlib.pyplot as plt

# Plot the distribution of plasmid sizes
plt.hist(plasmid_data['size'], bins=20)
plt.title('Distribution of Ti Plasmid Sizes')
plt.xlabel('Size (kb)')
plt.ylabel('Frequency')
plt.show()
```
