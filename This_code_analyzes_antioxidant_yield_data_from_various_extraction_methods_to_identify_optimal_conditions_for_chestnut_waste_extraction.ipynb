### Data Analysis of Antioxidant Yields from Extraction Methods

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for antioxidant yields from different extraction methods
extraction_methods = ['UAE', 'MAE', 'SFE', 'Solvent Extraction']
yields = [91.08, 85.5, 80.0, 75.0]  # hypothetical yield percentages

df = pd.DataFrame({'Extraction Method': extraction_methods, 'Yield (%)': yields})

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(df['Extraction Method'], df['Yield (%)'], color='skyblue')
plt.title('Antioxidant Yield from Different Extraction Methods')
plt.xlabel('Extraction Method')
plt.ylabel('Yield (%)')
plt.ylim(0, 100)
plt.grid(axis='y')
plt.show()
```
