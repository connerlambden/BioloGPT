This notebook analyzes the solubility of calcium carbonate at various pH levels.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for pH and solubility
data = {'pH': [6, 7, 8, 9, 10], 'solubility': [0.1, 0.2, 0.3, 0.25, 0.15]}
df = pd.DataFrame(data)

plt.plot(df['pH'], df['solubility'], marker='o')
plt.title('Calcium Carbonate Solubility vs pH')
plt.xlabel('pH')
plt.ylabel('Solubility (g/L)')
plt.grid()
plt.show()
```

The graph above illustrates the relationship between pH and the solubility of calcium carbonate.
