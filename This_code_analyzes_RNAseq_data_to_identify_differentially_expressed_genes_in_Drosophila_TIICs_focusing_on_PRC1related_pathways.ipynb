This notebook will analyze RNA-seq data from Drosophila TIICs to identify genes affected by PRC1 loss.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load RNA-seq data
data = pd.read_csv('path_to_rna_seq_data.csv')

# Perform differential expression analysis
results = stats.ttest_ind(data['control'], data['prc1_loss'])

# Output results
results_df = pd.DataFrame({'gene': data['gene'], 'p_value': results.pvalue})
results_df.to_csv('differential_expression_results.csv', index=False)
```
