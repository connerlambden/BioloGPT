### Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

### Step 2: Create a DataFrame with antimicrobial activity data
pathogens = ['S. aureus', 'P. aeruginosa', 'C. albicans', 'A. niger']
efficacy = [90, 85, 80, 75]
df = pd.DataFrame({'Pathogen': pathogens, 'Efficacy (%)': efficacy})

### Step 3: Plot the data
plt.bar(df['Pathogen'], df['Efficacy (%)'], color='blue')
plt.title('Antimicrobial Efficacy of rPET/Cu-Alg Nanofibers')
plt.xlabel('Pathogen')
plt.ylabel('Efficacy (%)')
plt.show()
