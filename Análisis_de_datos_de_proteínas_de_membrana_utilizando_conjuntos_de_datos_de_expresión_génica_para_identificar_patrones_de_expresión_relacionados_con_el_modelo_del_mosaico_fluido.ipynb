Análisis de expresión génica de proteínas de membrana.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('dataset_proteinas.csv')
sns.boxplot(x='tipo', y='expresion', data=data)
plt.title('Expresión de Proteínas de Membrana')
plt.show()
```

Este análisis permite visualizar la variabilidad en la expresión de diferentes tipos de proteínas de membrana.

```python
# Código para análisis de correlación entre proteínas y lípidos
data.corr().style.background_gradient(cmap='coolwarm')
```
