This notebook analyzes codon usage in mRNA sequences to identify optimal codons.

```python
import pandas as pd
# Load mRNA dataset
dataset = pd.read_csv('mRNA_sequences.csv')
# Analyze codon usage patterns
codon_usage = dataset['sequence'].apply(lambda x: count_codons(x))
```

The analysis provides insights into which codons are most frequently used in highly expressed genes.

```python
# Function to count codons

def count_codons(sequence):
    codons = [sequence[i:i+3] for i in range(0, len(sequence), 3)]
    return pd.Series(codons).value_counts()
```
