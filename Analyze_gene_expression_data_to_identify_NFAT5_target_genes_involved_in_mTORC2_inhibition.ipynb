This notebook will analyze gene expression datasets to find correlations between NFAT5 expression and mTORC2 activity.

```python
import pandas as pd
# Load gene expression data
gene_data = pd.read_csv('gene_expression_data.csv')
# Analyze NFAT5 and mTORC2 target genes
nfat5_targets = gene_data[gene_data['gene'] == 'NFAT5']
```

The analysis will focus on identifying significant changes in mTORC2 target genes in relation to NFAT5 expression.

```python
mTORC2_targets = gene_data[gene_data['gene'].isin(['AKT', 'RICTOR', 'SIN1'])]
# Correlation analysis
correlation = nfat5_targets.corrwith(mTORC2_targets)
```
