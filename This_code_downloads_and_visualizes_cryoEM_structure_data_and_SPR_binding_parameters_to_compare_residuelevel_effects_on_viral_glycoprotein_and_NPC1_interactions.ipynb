Below is a step-by-step Jupyter notebook code to download structure data from the PDB and visualize the qualitative effects of key residue changes on binding affinity using Plotly.

Next, we visualize the residue interaction effects using Plotly.

```python
import plotly.graph_objects as go

residues = ['Residue79', 'Residue141', 'Residue142', 'Residue148']
SUDV_effect = [1, 1, -1, 1]  # 1 indicates enhancing binding, -1 indicates reducing binding
EBOV_effect = [-1, -1, 1, -1]

fig = go.Figure(data=[go.Bar(name='SUDV', x=residues, y=SUDV_effect),
                         go.Bar(name='EBOV', x=residues, y=EBOV_effect)])

fig.update_layout(barmode='group', title='Residue Effects on Binding Affinity')
fig.show()
```
