This notebook will analyze the growth metrics of Artemia salina exposed to different concentrations of delorazepam.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Concentration (µg/L)': [1, 5, 10], 'Hatching Percentage': [75, 85, 90]}
df = pd.DataFrame(data)

plt.bar(df['Concentration (µg/L)'], df['Hatching Percentage'], color='blue')
plt.title('Hatching Percentage vs Delorazepam Concentration')
plt.xlabel('Concentration (µg/L)')
plt.ylabel('Hatching Percentage (%)')
plt.show()
```

The above code visualizes the relationship between delorazepam concentration and hatching percentage.
