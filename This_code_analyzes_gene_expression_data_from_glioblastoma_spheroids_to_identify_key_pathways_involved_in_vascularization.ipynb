This notebook will analyze gene expression data from glioblastoma spheroids to identify pathways related to vascularization.

```python
import pandas as pd
import numpy as np

# Load gene expression data
gene_data = pd.read_csv('glioblastoma_spheroids_expression.csv')

# Analyze pathways related to vascularization
vascular_pathways = gene_data[gene_data['pathway'].str.contains('vascularization')]

# Output results
vascular_pathways.to_csv('vascularization_analysis.csv', index=False)
```
