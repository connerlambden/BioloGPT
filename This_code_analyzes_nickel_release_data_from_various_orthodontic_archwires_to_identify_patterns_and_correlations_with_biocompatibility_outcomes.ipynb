### Nickel Release Analysis
This notebook analyzes the nickel release data from different orthodontic archwires, focusing on their biocompatibility implications.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for nickel release
archwire_data = {
    'Archwire Type': ['NiTi Superelastic', 'Heat-Activated NiTi', 'Stainless Steel CrNi', 'Epoxy-Coated NiTi'],
    'Nickel Release (µg/L)': [0.93, 0.67, 0.38, 0.25],
    'Recommended Duration (weeks)': [4, 8, 12, 6]
}

# Create DataFrame
df = pd.DataFrame(archwire_data)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['Archwire Type'], df['Nickel Release (µg/L)'], color='skyblue')
plt.title('Nickel Release from Orthodontic Archwires')
plt.xlabel('Archwire Type')
plt.ylabel('Nickel Release (µg/L)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
