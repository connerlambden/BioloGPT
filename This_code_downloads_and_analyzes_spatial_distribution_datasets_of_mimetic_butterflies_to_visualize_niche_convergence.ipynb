This notebook analyzes the spatial occurrence and niche data of Heliconiini and Ithomiini butterflies based on provided distribution maps and phenotypic classification datasets.

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('https://zenodo.org/record/10906853/files/occurrences.csv')
fig = px.scatter_mapbox(df, lat='latitude', lon='longitude', color='phenotypic_group', zoom=3, mapbox_style='open-street-map', title='Butterfly Mimicry Spatial Distribution')
fig.show()
```

The code downloads occurrence data from Zenodo, maps it with Plotly, and color codes points by phenotypic groups to illustrate mimicry rings and spatial niches.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Comparative analysis of climatic niche
climate_df = pd.read_csv('https://zenodo.org/record/10903661/files/climatic_data.csv')
sns.boxplot(x='tribe', y='temperature', data=climate_df)
plt.title('Comparative Climatic Niche by Tribe')
plt.show()
```
