This notebook analyzes gene expression data from 3D tumor models to identify pathways involved in tumor cell behavior.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_3D.csv')
sns.boxplot(data=data, x='gene', y='expression')
plt.title('Gene Expression in 3D Tumor Models')
plt.show()
```

The analysis reveals significant differences in gene expression between 2D and 3D cultures, highlighting pathways that may be targeted for therapy.
