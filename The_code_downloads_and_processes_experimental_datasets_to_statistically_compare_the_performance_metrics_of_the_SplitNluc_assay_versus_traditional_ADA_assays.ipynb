This notebook details the data extraction, normalization, and comparative statistical analysis of ADA detection performance using the SplitNluc platform.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('splitnluc_data.csv')
# Normalize and compare sensitivity
sensitivity = data[['Method', 'Sensitivity']]
sensitivity.groupby('Method').mean().plot(kind='bar', color='#6A0C76')
plt.title('Average Sensitivity Comparison')
plt.xlabel('Assay Method')
plt.ylabel('Sensitivity')
plt.show()
```

The above code block illustrates a basic statistical comparison of assay sensitivity, providing a visual bar chart to contrast the SplitNluc assay with conventional methods.

```python
# Additional code for advanced statistical tests
from scipy.stats import ttest_ind

method1 = data[data['Method'] == 'SplitNluc']['Sensitivity']
method2 = data[data['Method'] == 'ECLIA']['Sensitivity']
stat, pvalue = ttest_ind(method1, method2)
print('T-test statistic:', stat, 'P-value:', pvalue)
```

This statistical test evaluates the significance of the sensitivity difference observed between the two methods.
