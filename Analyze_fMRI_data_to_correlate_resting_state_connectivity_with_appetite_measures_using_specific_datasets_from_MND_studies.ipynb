This notebook will analyze fMRI data from MND patients to explore the relationship between resting state connectivity and subjective appetite measures.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load fMRI and appetite data
data = pd.read_csv('mnd_fmri_appetite_data.csv')

# Analyze correlations
correlation_matrix = data.corr()

# Visualize correlations
plt.figure(figsize=(10, 8))
plt.title('Correlation between Resting State Connectivity and Appetite Measures')
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.show()
```

The analysis will provide insights into how changes in brain connectivity relate to appetite in MND patients.
