### Rumen Microbiome Analysis with QIIME2

This notebook demonstrates the analysis of rumen microbiome sequencing data to identify shifts in bacterial taxa associated with CPS supplementation.

```python
# Import QIIME2
import qiime2
from qiime2 import Artifact

# Load the sequencing data
sequences = Artifact.load('path/to/sequences.qza')

# Perform diversity analysis
core_metrics = qiime2.plugins.diversity.methods.core_metrics_phylogenetic(
    sequences=sequences,
    sampling_depth=20000,
    phylogeny='path/to/phylogeny.qza'
)
```

### Differential Abundance Analysis

Identify bacterial genera that are significantly affected by CPS supplementation.

```python
# Load results
composition = core_metrics.emperor.plot

# Save results
composition.export('rumen_composition.html')
# Additional code for differential abundance using DESeq2 or similar methods
```
