### Analysis of Antiviral Activity of 1-Oxo-bisabolone
This notebook will analyze the antiviral activity of 1-oxo-bisabolone against HSV-1 and SARS-CoV-2, focusing on its potential synergistic effects with other antiviral agents.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for antiviral activity
data = {'Virus': ['HSV-1', 'SARS-CoV-2'], 'Inhibition (%)': [69, 65]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Virus'], df['Inhibition (%)'], color=['blue', 'orange'])
plt.title('Antiviral Activity of 1-Oxo-bisabolone')
plt.xlabel('Virus')
plt.ylabel('Inhibition (%)')
plt.show()
```
