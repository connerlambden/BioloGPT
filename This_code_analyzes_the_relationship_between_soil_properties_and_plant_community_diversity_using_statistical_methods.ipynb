### Data Analysis of Plant Community and Soil Properties
This notebook will analyze the collected data on plant community characteristics and soil factors.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('alpine_grassland_data.csv')

# Analyze the correlation between soil properties and plant diversity
correlation = data.corr()
sns.heatmap(correlation, annot=True)
plt.title('Correlation between Soil Properties and Plant Diversity')
plt.show()
```

### Statistical Analysis
Conduct statistical tests to determine the significance of differences in plant diversity across slope aspects.

```python
from scipy import stats

# Perform ANOVA
f_statistic, p_value = stats.f_oneway(data['diversity_shady'], data['diversity_sunny'])
print(f'F-statistic: {f_statistic}, P-value: {p_value}')
```
