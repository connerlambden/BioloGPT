### Data Analysis of T. evansi Genetic Sequences
This notebook will analyze genetic sequences from T. evansi isolates to identify variations that may correlate with disease severity.

```python
import pandas as pd
# Load genetic sequence data
sequences = pd.read_csv('trypanosoma_sequences.csv')
# Analyze variations
variations = sequences['sequence'].apply(lambda x: len(set(x)))
variations.describe()
```

### Results
The analysis will provide insights into the genetic diversity of T. evansi isolates.

```python
# Display results
variations.plot(kind='hist')
```
