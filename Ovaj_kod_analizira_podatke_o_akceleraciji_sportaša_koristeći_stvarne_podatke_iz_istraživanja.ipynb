Analiza podataka o akceleraciji sportaša.

```python
import pandas as pd

# Učitavanje podataka o akceleraciji
data = pd.read_csv('acceleration_data.csv')

# Analiza podataka
mean_acceleration = data['acceleration'].mean()
print(f'Mean Acceleration: {mean_acceleration} m/s²')
```

Ovaj kod izračunava prosječnu akceleraciju sportaša na temelju prikupljenih podataka.

```python
# Vizualizacija podataka
import matplotlib.pyplot as plt
plt.plot(data['time'], data['acceleration'])
plt.title('Acceleration over Time')
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s²)')
plt.show()
```
