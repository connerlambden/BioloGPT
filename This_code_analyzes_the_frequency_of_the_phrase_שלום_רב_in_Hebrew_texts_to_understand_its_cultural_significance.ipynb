This notebook will analyze Hebrew texts to quantify the usage of 'שלום רב' and its variations.

```python
import pandas as pd

# Load Hebrew text data
data = pd.read_csv('hebrew_texts.csv')

# Count occurrences of 'שלום רב'
count = data['text'].str.count('שלום רב').sum()
print(f"The phrase 'שלום רב' appears {count} times in the dataset.")
```

The analysis will provide insights into how frequently 'שלום רב' is used in different contexts.

```python
# Visualize the results
import matplotlib.pyplot as plt

plt.bar(['שלום רב'], [count])
plt.title('Frequency of the phrase שלום רב')
plt.ylabel('Count')
plt.show()
```
