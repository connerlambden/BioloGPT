This notebook will analyze gene expression data from broiler chickens exposed to AFB1, focusing on CYP2A6 and Nrf2.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_data.csv')
sns.boxplot(x='treatment', y='expression', data=data)
plt.title('Gene Expression of CYP2A6 and Nrf2')
plt.show()
```
