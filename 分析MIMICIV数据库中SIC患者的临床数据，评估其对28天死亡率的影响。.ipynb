### 数据准备与分析
本节将使用MIMIC-IV数据库中的数据，分析脓毒症诱发凝血病（SIC）患者的临床特征与28天死亡率的关系。

```python
import pandas as pd

# 读取MIMIC-IV数据库中的数据
df = pd.read_csv('mimic_iv_data.csv')

# 数据清洗与预处理
df_cleaned = df.dropna(subset=['SIC', 'mortality_28_days'])

# 统计分析
mortality_rate = df_cleaned.groupby('SIC')['mortality_28_days'].mean()
print(mortality_rate)
```
