### Genetic Analysis of Homozygosity
This notebook analyzes genetic data to identify homozygous and heterozygous variants in a population.

```python
import pandas as pd

data = pd.read_csv('genetic_data.csv')

homozygous = data[data['allele1'] == data['allele2']]
heterozygous = data[data['allele1'] != data['allele2']]

homozygous.to_csv('homozygous_variants.csv')
heterozygous.to_csv('heterozygous_variants.csv')
```
