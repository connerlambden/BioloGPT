This notebook will analyze the peptide sequences identified in Podolica milk, focusing on their bioactivity and potential health benefits.

```python
import pandas as pd

# Load peptide data
peptide_data = pd.read_csv('podolica_peptides.csv')

# Analyze bioactivity
bioactivity_analysis = peptide_data.groupby('function').count()

# Display results
bioactivity_analysis
```

The analysis provides insights into the distribution of bioactive functions among the peptides identified in Podolica milk.

```python
# Visualize the distribution of bioactive functions
import matplotlib.pyplot as plt

bioactivity_analysis.plot(kind='bar')
plt.title('Distribution of Bioactive Functions in Podolica Milk Peptides')
plt.xlabel('Function')
plt.ylabel('Count')
plt.show()
```
