Below we detail the steps for downloading imaging datasets, cleaning the data, and performing statistical analysis to compare report quality metrics in structured and free-text reports.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Load dataset containing report quality metrics: decision-making support, linguistic quality, overall quality
# Assume dataset.csv contains columns: ReportType, DecisionSupport, LinguisticQuality, OverallQuality

data = pd.read_csv('dataset.csv')

# Separate datasets
sr_data = data[data['ReportType'] == 'SR']
ftr_data = data[data['ReportType'] == 'FTR']

# Perform t-tests
dec_support_ttest = stats.ttest_ind(sr_data['DecisionSupport'], ftr_data['DecisionSupport'])
ling_quality_ttest = stats.ttest_ind(sr_data['LinguisticQuality'], ftr_data['LinguisticQuality'])
overall_quality_ttest = stats.ttest_ind(sr_data['OverallQuality'], ftr_data['OverallQuality'])

print('Decision Support T-test:', dec_support_ttest)
print('Linguistic Quality T-test:', ling_quality_ttest)
print('Overall Quality T-test:', overall_quality_ttest)

# Plot comparison
plt.figure(figsize=(8,6))
plt.boxplot([ftr_data['OverallQuality'], sr_data['OverallQuality']], labels=['FTR', 'SR'])
plt.title('Overall Report Quality Comparison')
plt.ylabel('Quality Score')
plt.show()
```
