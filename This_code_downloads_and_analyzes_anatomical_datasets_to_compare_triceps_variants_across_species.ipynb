This notebook imports anatomical datasets and metadata relevant to triceps muscle variations to perform statistical and comparative analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load anatomical variation dataset
anatomy_df = pd.read_csv('anatomical_variations.csv')
# Filter for triceps data
triceps_df = anatomy_df[anatomy_df['muscle'] == 'triceps brachii']
# Plot distribution of number of heads
plt.figure(figsize=(10,6))
plt.hist(triceps_df['number_of_heads'], bins=range(3,8), color='#6A0C76', edgecolor='black')
plt.title('Distribution of Triceps Brachii Heads')
plt.xlabel('Number of Heads')
plt.ylabel('Frequency')
plt.show()
```

This visualization helps in understanding the prevalence of triceps head variations across species.
