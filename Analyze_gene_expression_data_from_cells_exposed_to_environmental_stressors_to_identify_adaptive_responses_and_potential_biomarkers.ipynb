This notebook analyzes gene expression data to identify changes associated with environmental stress exposure.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_data.csv')
sns.boxplot(data=data, x='Condition', y='Expression')
plt.title('Gene Expression under Stress Conditions')
plt.show()
```
