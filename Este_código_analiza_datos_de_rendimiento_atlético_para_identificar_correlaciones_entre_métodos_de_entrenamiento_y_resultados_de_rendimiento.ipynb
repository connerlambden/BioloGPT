Análisis de datos de rendimiento atlético.

```python
import pandas as pd

# Cargar datos de rendimiento
data = pd.read_csv('performance_data.csv')

# Análisis de correlación
data.corr()
```

Este análisis permite identificar qué métodos de entrenamiento están más correlacionados con mejoras en el rendimiento.

```python
# Visualización de correlaciones
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(data.corr(), annot=True)
plt.show()
```
