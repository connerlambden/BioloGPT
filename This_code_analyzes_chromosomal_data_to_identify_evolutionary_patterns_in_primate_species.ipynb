### Chromosomal Data Analysis
This notebook will analyze the chromosomal data of humans and other primates to explore evolutionary relationships.

```python
import pandas as pd

# Load chromosomal data
chromosome_data = pd.read_csv('chromosome_data.csv')

# Analyze the number of chromosomes
chromosome_counts = chromosome_data.groupby('species')['chromosome_count'].sum()
print(chromosome_counts)
```
