This notebook will analyze RNA-seq data from BHK-7-FC cells to identify genes that are differentially expressed following 12C6 heavy ion irradiation.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load RNA-seq data
rna_seq_data = pd.read_csv('path_to_rna_seq_data.csv')

# Perform differential expression analysis
results = stats.ttest_ind(rna_seq_data['irradiated'], rna_seq_data['control'])

# Output results
print(results)
```
