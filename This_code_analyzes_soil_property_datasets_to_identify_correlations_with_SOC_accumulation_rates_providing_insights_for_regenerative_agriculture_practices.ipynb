### Soil Property and SOC Analysis
This notebook analyzes the relationship between local soil properties and SOC accumulation using provided datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
soil_data = pd.read_csv('soil_properties.csv')
soc_data = pd.read_csv('soc_accumulation.csv')

# Merge datasets on relevant keys
merged_data = pd.merge(soil_data, soc_data, on='location')

# Analyze correlations
correlation = merged_data.corr()['SOC_accumulation']

# Plot results
plt.figure(figsize=(10, 6))
plt.bar(correlation.index, correlation.values)
plt.title('Correlation of Soil Properties with SOC Accumulation')
plt.xlabel('Soil Properties')
plt.ylabel('Correlation Coefficient')
plt.show()
```
