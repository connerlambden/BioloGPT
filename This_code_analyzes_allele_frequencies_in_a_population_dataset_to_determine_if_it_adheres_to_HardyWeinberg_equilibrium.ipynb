### Population Genetics Analysis
This notebook will analyze allele frequencies from a population dataset to check for Hardy-Weinberg equilibrium.

```python
import pandas as pd

# Load population data
data = pd.read_csv('population_data.csv')

# Calculate allele frequencies
allele_A = data['A'].sum() / data['total']
allele_a = data['a'].sum() / data['total']

# Check Hardy-Weinberg equilibrium
expected_AA = allele_A**2
expected_Aa = 2 * allele_A * allele_a
expected_aa = allele_a**2

print(f'Expected frequencies: AA: {expected_AA}, Aa: {expected_Aa}, aa: {expected_aa}')
```

The results will indicate whether the observed genotype frequencies align with the expected frequencies under Hardy-Weinberg equilibrium.
