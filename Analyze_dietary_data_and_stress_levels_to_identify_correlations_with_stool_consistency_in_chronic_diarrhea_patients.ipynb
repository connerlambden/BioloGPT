This notebook will analyze dietary intake and stress levels in patients with chronic diarrhea to identify potential correlations.

```python
import pandas as pd

diet_data = pd.read_csv('diet_data.csv')
stress_data = pd.read_csv('stress_data.csv')

# Merge datasets on patient ID
merged_data = pd.merge(diet_data, stress_data, on='patient_id')

# Analyze correlations
correlation = merged_data.corr()
```

The analysis will provide insights into how dietary choices and stress levels may influence stool consistency.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize correlations
sns.heatmap(correlation, annot=True)
plt.title('Correlation between Diet, Stress, and Stool Consistency')
plt.show()
```
