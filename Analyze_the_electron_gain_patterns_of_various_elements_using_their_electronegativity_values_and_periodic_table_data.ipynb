This notebook will analyze the electronegativity of elements that gain electrons, focusing on halogens and oxygen.

```python
import pandas as pd

elements = {'Element': ['Fluorine', 'Chlorine', 'Oxygen', 'Bromine'], 'Electronegativity': [3.98, 3.16, 3.44, 2.96]}
df = pd.DataFrame(elements)
df.sort_values(by='Electronegativity', ascending=False) # Sort by electronegativity
```

The sorted DataFrame will show the elements ranked by their electronegativity, indicating their tendency to gain electrons.

```python
df.plot(kind='bar', x='Element', y='Electronegativity', title='Electronegativity of Electron-Gaining Elements')
```
