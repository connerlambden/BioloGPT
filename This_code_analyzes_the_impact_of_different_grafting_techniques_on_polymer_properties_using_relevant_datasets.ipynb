### Polymer Grafting Analysis
This notebook analyzes the effects of grafting on various polymers using data from recent studies.

```python
import pandas as pd

# Load relevant datasets
datasets = {
    'PHB': {'thermal_stability': 9, 'degradability': 36},
    'Chitosan': {'antioxidant_activity': 1464.2},
    'Cellulose': {'mechanical_properties': 1.8}
}

# Create a DataFrame
polymer_df = pd.DataFrame(datasets)
polymer_df.T
```

### Summary of Grafting Effects
The analysis shows significant improvements in thermal stability, antioxidant activity, and mechanical properties due to grafting.

```python
polymer_df.T.plot(kind='bar', title='Grafting Effects on Polymer Properties')
```
