This notebook will analyze GPS tracking data of species to identify patterns of spatial and temporal resource utilization.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load species activity data
data = pd.read_csv('species_activity_data.csv')

# Analyze activity patterns
activity_patterns = data.groupby(['species', 'time']).mean()

# Visualize results
plt.figure(figsize=(10,6))
for species in activity_patterns.index.levels[0]:
    plt.plot(activity_patterns.loc[species].index, activity_patterns.loc[species]['activity_level'], label=species)
plt.title('Species Activity Patterns')
plt.xlabel('Time of Day')
plt.ylabel('Activity Level')
plt.legend()
plt.show()
```

This analysis will help identify how different species adjust their activity patterns in response to resource availability and competition.
