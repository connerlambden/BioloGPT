Generate a detailed notebook to preprocess MRI workflow datasets and compute statistical differences between optimized and reference facilities.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Code to load and analyze MRI workflow data from provided datasets
mri_data = pd.read_csv('mri_workflow_data.csv')
# Compute mean turnaround times
mean_times = mri_data.groupby('facility')['turnaround_time'].mean()
print(mean_times)
plt.bar(mean_times.index, mean_times.values)
plt.title('Mean Table Turnaround Time')
plt.xlabel('Facility')
plt.ylabel('Time (min)')
plt.show()
```

The notebook includes steps for data cleaning, statistical tests (t-tests), and visualization of workflow distributions.

```python
# Further analysis and visualizations here
import seaborn as sns
sns.histplot(data=mri_data[mri_data['facility'] == 'Optimized'], x='turnaround_time', kde=True, color='purple', label='Optimized')
sns.histplot(data=mri_data[mri_data['facility'] == 'Reference'], x='turnaround_time', kde=True, color='orange', label='Reference')
plt.legend()
plt.title('Turnaround Time Distribution')
plt.show()
```
