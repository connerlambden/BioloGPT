Download and preprocess GB cell line and patient RNA-seq datasets to compare mitochondrial biogenesis markers in AMPK-intact vs AMPK-inhibited conditions.

```python
import pandas as pd
import numpy as np
# Load RNA-seq expression data
expression_data = pd.read_csv('gb_expression.csv')
# Filter for mitochondrial biogenesis markers
markers = ['MTCO1', 'MTCYB', 'ATP5A']
marker_expression = expression_data[markers]
# Compute average expression for AMPK high vs low groups
grouped = marker_expression.groupby(expression_data['AMPK_status']).mean()
print(grouped)
```
