### Performance Analysis of GDPS Designs
This notebook analyzes the performance metrics of secondary-side-regulated (SSR) and primary-side-regulated (PSR) gate drive power supplies.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for performance metrics
data = {
    'Design': ['SSR', 'PSR'],
    'Inter-Winding Capacitance (pF)': [150, 100],
    'Load-Voltage-Regulation-Rate (%)': [5, 2],
    'Conversion Efficiency (%)': [90, 95],
    'Hardware Complexity': [3, 1]  # 1-5 scale
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting the performance metrics
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# Inter-Winding Capacitance
ax[0, 0].bar(df['Design'], df['Inter-Winding Capacitance (pF)'], color='blue')
ax[0, 0].set_title('Inter-Winding Capacitance')
ax[0, 0].set_ylabel('Capacitance (pF)')

# Load-Voltage-Regulation-Rate
ax[0, 1].bar(df['Design'], df['Load-Voltage-Regulation-Rate (%)'], color='orange')
ax[0, 1].set_title('Load Voltage Regulation Rate')
ax[0, 1].set_ylabel('Regulation Rate (%)')

# Conversion Efficiency
ax[1, 0].bar(df['Design'], df['Conversion Efficiency (%)'], color='green')
ax[1, 0].set_title('Conversion Efficiency')
ax[1, 0].set_ylabel('Efficiency (%)')

# Hardware Complexity
ax[1, 1].bar(df['Design'], df['Hardware Complexity'], color='red')
ax[1, 1].set_title('Hardware Complexity')
ax[1, 1].set_ylabel('Complexity (1-5 scale)')

plt.tight_layout()
plt.show()
```

### Discussion
This analysis provides insights into the performance differences between SSR and PSR designs, highlighting the advantages of PSR in high-voltage applications.
