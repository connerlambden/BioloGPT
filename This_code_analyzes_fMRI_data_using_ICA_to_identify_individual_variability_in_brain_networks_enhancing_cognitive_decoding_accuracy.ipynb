### Step 1: Import Necessary Libraries
Import libraries for data manipulation and ICA analysis.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import FastICA
from sklearn.preprocessing import StandardScaler
```

### Step 2: Load fMRI Data
Load the fMRI dataset for analysis.

```python
# Load your fMRI data here
# Example: data = pd.read_csv('fmri_data.csv')
# For demonstration, we will create a random dataset
np.random.seed(0)
data = np.random.rand(100, 50)  # 100 samples, 50 features
```

### Step 3: Preprocess Data
Standardize the data before applying ICA.

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

### Step 4: Apply ICA
Perform ICA to extract independent components.

```python
ica = FastICA(n_components=10, random_state=0)
components = ica.fit_transform(data_scaled)  # Extracted components
```
