This notebook will analyze student performance data to evaluate the effectiveness of active learning strategies.

```python
import pandas as pd

# Load student performance data
data = pd.read_csv('student_performance.csv')

# Analyze the impact of online study questions on exam scores
results = data.groupby('access_to_study_questions')['exam_score'].mean()
```

The analysis will provide insights into how access to study questions correlates with exam performance.

```python
results.plot(kind='bar', title='Average Exam Scores by Access to Study Questions')
```
