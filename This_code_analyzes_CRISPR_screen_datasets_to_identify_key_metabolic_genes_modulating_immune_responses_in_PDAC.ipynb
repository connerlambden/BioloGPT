Loading and preprocessing the CRISPR screen dataset for the pancreatic cancer model.

```python
import pandas as pd
# Load CRISPR screen data
screen_data = pd.read_csv('crisp_screen_data.csv')
# Filter for significant gene hits
significant_hits = screen_data[screen_data['p_value'] < 0.05]
print(significant_hits.head())
```

Visualize the distribution of sgRNA counts for PCCB and other significant hits.

```python
import plotly.express as px
fig = px.histogram(significant_hits, x='sgRNA_count', nbins=30, title='sgRNA Count Distribution')
fig.show()
```
