This notebook will analyze gene expression data from immune cells to identify patterns related to TCR and MHC interactions.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('immune_response_data.csv')
sns.boxplot(x='TCR_signaling', y='Gene_expression', data=data)
plt.title('TCR Signaling vs Gene Expression')
plt.show()
```

The analysis reveals how TCR signaling influences gene expression in immune cells.
