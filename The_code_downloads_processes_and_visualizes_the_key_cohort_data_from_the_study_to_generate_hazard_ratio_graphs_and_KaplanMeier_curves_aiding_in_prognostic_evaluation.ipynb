This section details the steps to load and analyze the HUS cohort data, subsequently visualizing the association between PI-RADS scores and PCSM.

```python
import pandas as pd
import plotly.express as px

# Assuming the dataset is loaded from a CSV file
data = pd.read_csv('hus_cohort_data.csv')
# Generate a bar plot for hazard ratios
fig = px.bar(data, x='PI-RADS', y='Hazard_Ratio', title='Hazard Ratio for PCSM by PI-RADS Score')
fig.show()
```

The notebook includes detailed explanations to help users understand the data processing and visualization pipeline.

```python
# Additional analysis: Kaplan-Meier survival curve
import lifelines as lf
from lifelines import KaplanMeierFitter

kmf = KaplanMeierFitter()
# Fit the model for patients with PI-RADS score 5
mask = data['PI-RADS'] == 5
data_score5 = data[mask]
kmf.fit(data_score5['time_to_event'], event_observed=data_score5['event'])
kmf.plot(title='KM Curve for PI-RADS Score 5 Patients')
```
