### Chloride Concentration Analysis
This notebook analyzes chloride concentration data from multiple water samples.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Sample': ['Sample 1', 'Sample 2', 'Sample 3', 'Sample 4'], 'Chloride Concentration (mg/L)': [100, 150, 200, 250]}
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Sample'], df['Chloride Concentration (mg/L)'])
plt.title('Chloride Concentration in Water Samples')
plt.xlabel('Water Samples')
plt.ylabel('Chloride Concentration (mg/L)')
plt.show()
```
