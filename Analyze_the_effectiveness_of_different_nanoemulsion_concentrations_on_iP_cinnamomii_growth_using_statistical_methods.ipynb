This notebook analyzes the effectiveness of various nanoemulsion concentrations on the growth of <i>P. cinnamomi</i>.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for nanoemulsion concentrations and growth inhibition
data = {'Concentration (ppm)': [0, 100, 200, 300, 400, 500], 'Inhibition (%)': [0, 30, 60, 80, 90, 95]}
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(df['Concentration (ppm)'], df['Inhibition (%)'], marker='o')
plt.title('Effect of Nanoemulsion Concentration on P. cinnamomi Growth Inhibition')
plt.xlabel('Concentration (ppm)')
plt.ylabel('Inhibition (%)')
plt.grid()
plt.show()
```

This plot illustrates the relationship between nanoemulsion concentration and the inhibition of <i>P. cinnamomi</i> growth.
