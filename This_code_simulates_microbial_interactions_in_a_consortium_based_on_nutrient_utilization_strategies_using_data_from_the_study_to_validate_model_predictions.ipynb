### Simulation of Microbial Consortium Interactions
This notebook simulates the interactions among different microbial species in a consortium, focusing on nutrient utilization and stability.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the simulation
num_species = 3
nutrient_levels = np.linspace(0, 10, 100)

# Simulate growth rates based on nutrient levels
growth_rates = np.array([0.1 * nutrient_levels, 0.2 * nutrient_levels, 0.15 * nutrient_levels])

# Plotting the growth rates
plt.figure(figsize=(10, 6))
for i in range(num_species):
    plt.plot(nutrient_levels, growth_rates[i], label=f'Species {i+1}')
plt.title('Growth Rates of Microbial Species vs Nutrient Levels')
plt.xlabel('Nutrient Levels')
plt.ylabel('Growth Rate')
plt.legend()
plt.grid(True)
plt.show()
```

### Discussion
This simulation illustrates how different species may respond to varying nutrient levels, highlighting the importance of ecological matching in maintaining stability.
