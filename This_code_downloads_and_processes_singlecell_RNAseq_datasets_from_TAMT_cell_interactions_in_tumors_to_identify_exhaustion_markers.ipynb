Overview of TAM-T cell interaction single-cell analysis.

```python
import scanpy as sc
adata = sc.read_h5ad('path_to_dataset.h5ad')
sc.pp.filter_genes(adata, min_cells=3)
sc.tl.pca(adata)
sc.pl.pca_variance_ratio(adata, log=True)
```

Differential expression analysis to uncover exhaustion markers.
