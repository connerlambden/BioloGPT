The notebook loads raw spectral imaging, FCS data, and applies statistical models to correlate lipid order (λMAX) with diffusion coefficients.

```python
import pandas as pd
import numpy as np
import plotly.express as px
# Download dataset from provided URLs (assumed available in the environment)
data = pd.read_csv('spectral_STED_data.csv')
# Calculate correlation between λMAX shift and diffusion metrics
correlation = data['lambda_max'].corr(data['diffusion_coefficient'])
fig = px.scatter(data, x='lambda_max', y='diffusion_coefficient', title='Correlation between Lipid Order and Diffusion')
fig.show()
print('Correlation coefficient:', correlation)
```
