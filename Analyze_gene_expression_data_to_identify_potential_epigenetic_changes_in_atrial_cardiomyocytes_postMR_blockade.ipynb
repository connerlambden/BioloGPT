This notebook will analyze RNA-seq data from atrial cardiomyocytes to identify differentially expressed genes following MR blockade.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load RNA-seq data
data = pd.read_csv('atrial_cardiomyocytes_rnaseq.csv')

# Perform differential expression analysis
results = stats.ttest_ind(data['control'], data['MR_blockade'])

# Output results
results
```
