This notebook details the steps to load and filter RNA-seq data, perform differential expression analysis, and visualize correlations among PDPN, CCR7, and EMT signature genes.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load RNA-seq data
data = pd.read_csv('GSE-x.csv')

# Filter samples with high PDPN and CCR7 expression
threshold = data['PDPN'].quantile(0.75)
filtered = data[(data['PDPN'] > threshold) & (data['CCR7'] > threshold)]

# Describe expression levels
print(filtered[['PDPN', 'CCR7', 'EMT_score']].describe())

# Plot histogram of PDPN expression
plt.hist(filtered['PDPN'], color='#6A0C76')
plt.title('PDPN Expression Distribution')
plt.xlabel('Expression Level')
plt.ylabel('Frequency')
plt.show()
```

Next, visualize the correlation between PDPN and EMT scores.

```python
import seaborn as sns
sns.scatterplot(x='PDPN', y='EMT_score', data=filtered, color='#6A0C76')
plt.title('Correlation of PDPN and EMT Score')
plt.xlabel('PDPN Expression')
plt.ylabel('EMT Score')
plt.show()
```
