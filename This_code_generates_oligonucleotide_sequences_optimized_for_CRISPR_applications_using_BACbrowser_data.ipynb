### Step 1: Import Required Libraries
Import necessary libraries for data handling and oligonucleotide design.

```python
import pandas as pd
import numpy as np

# Load BAC-browser data for oligonucleotide design
bac_data = pd.read_csv('bac_browser_data.csv')
```

### Step 2: Define Oligonucleotide Design Function
Create a function to design oligonucleotides based on user-defined parameters.

```python
def design_oligos(sequence, oligo_length=40, overlap=20):
    oligos = []
    for i in range(0, len(sequence), oligo_length - overlap):
        oligo = sequence[i:i + oligo_length]
        if len(oligo) == oligo_length:
            oligos.append(oligo)
    return oligos
```

### Step 3: Generate Oligonucleotides
Use the defined function to generate oligonucleotides from a sample sequence.

```python
sample_sequence = 'ATGCGTACGTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAGCTAG'
oligos = design_oligos(sample_sequence)
print(oligos)
```
