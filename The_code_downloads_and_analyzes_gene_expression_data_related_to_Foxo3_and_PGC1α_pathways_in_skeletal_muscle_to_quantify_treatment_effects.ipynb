This notebook loads gene expression datasets from treated and control groups to compare the modulation of key muscle atrophy markers.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example: Load expression data
control = pd.read_csv('control_group.csv')
treatment = pd.read_csv('treatment_group.csv')

data = pd.merge(control, treatment, on='gene', suffixes=('_control', '_treatment'))

# Plot differences in Foxo3 and PGC-1α expression
genes_of_interest = ['Foxo3', 'PGC1A']
subset = data[data['gene'].isin(genes_of_interest)]

plt.bar(subset['gene'], subset['expression_treatment'] - subset['expression_control'], color='#6A0C76')
plt.xlabel('Gene')
plt.ylabel('Expression Difference')
plt.title('Treatment Effect on Muscle Atrophy Markers')
plt.show()
```

The code demonstrates how to visualize the differences in gene expression between treatment and control groups for key targets.
