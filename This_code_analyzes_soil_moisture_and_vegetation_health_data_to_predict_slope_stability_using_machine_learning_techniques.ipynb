### Data Preparation
Load datasets on soil moisture and vegetation indices.

```python
import pandas as pd
soil_data = pd.read_csv('soil_moisture_data.csv')
veg_data = pd.read_csv('vegetation_health_data.csv')
```

### Data Analysis
Merge datasets and analyze correlations.

```python
merged_data = pd.merge(soil_data, veg_data, on='location')
correlation = merged_data.corr()
```

### Model Training
Train a machine learning model to predict slope stability.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
X = merged_data[['soil_moisture', 'vegetation_index']]
y = merged_data['slope_stability']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = RandomForestRegressor()
model.fit(X_train, y_train)
```
