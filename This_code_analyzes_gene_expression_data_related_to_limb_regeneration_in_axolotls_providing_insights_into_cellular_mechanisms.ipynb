This notebook will analyze gene expression data from axolotl limb regeneration studies.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('axolotl_regeneration_data.csv')
sns.boxplot(x='treatment', y='gene_expression', data=data)
plt.title('Gene Expression in Regeneration Studies')
plt.show()
```

The analysis visualizes the differences in gene expression between treated and control groups, highlighting key genes involved in limb regeneration.
