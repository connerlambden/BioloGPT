```python
import pandas as pd
import plotly.express as px
# Sample dataset; real data to be retrieved from supplementary files
data = {'Condition': ['Light', 'Dark'], 'Speed': [0.41, 0.10]}
df = pd.DataFrame(data)
fig = px.bar(df, x='Condition', y='Speed', title='Coral Larvae Swimming Speeds')
fig.show()
```
