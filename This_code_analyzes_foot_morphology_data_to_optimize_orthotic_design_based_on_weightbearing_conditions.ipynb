### Foot Morphology Analysis
This notebook analyzes foot morphology data to determine optimal orthotic design parameters based on weight-bearing conditions.

```python
import pandas as pd

# Load foot morphology data
data = pd.read_csv('foot_morphology.csv')

# Analyze differences in morphology based on weight-bearing conditions
nwb_data = data[data['condition'] == 'NWB']
pwb_data = data[data['condition'] == 'PWB']

# Calculate mean differences
mean_diff = nwb_data.mean() - pwb_data.mean()
mean_diff
```

### Results
The analysis provides insights into how weight-bearing conditions affect foot morphology, guiding orthotic design.

```python
# Visualize results
import matplotlib.pyplot as plt

plt.bar(['NWB', 'PWB'], [nwb_data.mean(), pwb_data.mean()])
plt.title('Mean Foot Morphology by Weight-Bearing Condition')
plt.ylabel('Mean Measurements')
plt.show()
```
