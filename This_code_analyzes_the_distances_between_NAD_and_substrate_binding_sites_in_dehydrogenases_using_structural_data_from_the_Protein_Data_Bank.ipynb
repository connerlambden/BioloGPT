This notebook will analyze the distances between NAD and substrate binding sites in various dehydrogenases using structural data from the Protein Data Bank.

```python
import pandas as pd
import numpy as np

# Load structural data from PDB
pdb_data = pd.read_csv('pdb_structures.csv')

# Calculate distances between NAD and substrate binding sites
# (Assuming the data contains columns for NAD and substrate coordinates)
# Example calculation
pdb_data['distance'] = np.sqrt((pdb_data['NAD_x'] - pdb_data['substrate_x'])**2 + (pdb_data['NAD_y'] - pdb_data['substrate_y'])**2 + (pdb_data['NAD_z'] - pdb_data['substrate_z'])**2)

# Output the results
pdb_data[['PDB_ID', 'distance']].to_csv('NAD_substrate_distances.csv', index=False)
```
