Analyse des données transcriptomiques pour explorer les différences d'expression génique entre les spécialités en neurosciences.

```python
import pandas as pd

data = pd.read_csv('neuroscience_data.csv')
# Analyse des différences d'expression génique
results = data.groupby('specialty').mean()
```

Visualisation des résultats pour identifier les spécialités avec des différences significatives.

```python
import matplotlib.pyplot as plt

plt.bar(results.index, results['expression'])
plt.title('Différences d expression génique par spécialité')
plt.show()
```
