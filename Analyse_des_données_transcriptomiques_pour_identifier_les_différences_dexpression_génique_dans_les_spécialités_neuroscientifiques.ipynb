Analyse des données transcriptomiques pour explorer les différences d'expression génique entre les spécialités en neurosciences.

```python
import pandas as pd
# Charger les données transcriptomiques
transcriptomic_data = pd.read_csv('data.csv')
# Analyse des différences d'expression génique
results = transcriptomic_data.groupby('specialty').mean()
```

Les résultats montrent les niveaux d'expression génique moyens par spécialité.

```python
results.to_csv('results.csv')
```
