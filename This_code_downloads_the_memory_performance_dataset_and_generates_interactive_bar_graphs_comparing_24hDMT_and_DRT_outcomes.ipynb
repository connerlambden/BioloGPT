This notebook retrieves marmoset memory test performance data and visualizes task differences using Plotly.

```python
import plotly.graph_objects as go

# Data for memory tasks
tasks = ['24h-DMT', 'DRT']
performance = [90, 50]  # Example performance percentages

fig = go.Figure(data=[go.Bar(x=tasks, y=performance, marker_color='#6A0C76')])

fig.update_layout(title='Marmoset Memory Task Performance', xaxis_title='Memory Task', yaxis_title='Performance (%)')
fig.show()
```

The graph visualizes the performance differential, supporting the conclusion that task design impacts measured memory.
