### Analysis of Antibiotic Resistance Genes in Staphylococcus aureus
This notebook will analyze the presence of antibiotic resistance genes in different clonal complexes of Staphylococcus aureus using genomic data.

```python
import pandas as pd

# Load relevant datasets
ca_mrsa_data = pd.read_csv('path_to_ca_mrsa_data.csv')
ha_mrsa_data = pd.read_csv('path_to_ha_mrsa_data.csv')

# Analyze resistance patterns
ca_mrsa_resistance = ca_mrsa_data['resistance'].value_counts()
ha_mrsa_resistance = ha_mrsa_data['resistance'].value_counts()

# Visualize the results
import matplotlib.pyplot as plt
plt.bar(ca_mrsa_resistance.index, ca_mrsa_resistance.values, label='CA-MRSA')
plt.bar(ha_mrsa_resistance.index, ha_mrsa_resistance.values, label='HA-MRSA', alpha=0.5)
plt.xlabel('Resistance Type')
plt.ylabel('Frequency')
plt.title('Antibiotic Resistance in CA-MRSA vs HA-MRSA')
plt.legend()
plt.show()
```
