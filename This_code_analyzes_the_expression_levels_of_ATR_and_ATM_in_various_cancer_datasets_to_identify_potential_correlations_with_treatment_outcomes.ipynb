# Analysis of ATR and ATM Expression in Cancer Datasets
This notebook will explore the expression levels of ATR and ATM in different cancer types using publicly available datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load cancer expression data
cancer_data = pd.read_csv('cancer_expression_data.csv')

# Analyze ATR and ATM expression levels
atr_expression = cancer_data['ATR']
atm_expression = cancer_data['ATM']

# Plotting the expression levels
plt.figure(figsize=(10, 6))
plt.scatter(atr_expression, atm_expression)
plt.title('ATR vs ATM Expression in Cancer')
plt.xlabel('ATR Expression')
plt.ylabel('ATM Expression')
plt.grid(True)
plt.show()
```
