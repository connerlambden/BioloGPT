This notebook will analyze hemodynamic parameters from CFD simulations to identify correlations with aneurysm rupture risks.

```python
import pandas as pd
# Load hemodynamic data
data = pd.read_csv('hemodynamic_data.csv')
# Analyze correlations
correlations = data.corr()
```

The analysis will provide insights into which hemodynamic parameters are most predictive of aneurysm rupture.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(correlations, annot=True)
plt.title('Correlation Matrix of Hemodynamic Parameters')
plt.show()
```
