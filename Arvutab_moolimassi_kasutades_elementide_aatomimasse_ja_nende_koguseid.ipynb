Arvuta moolimass, kasutades järgmisi andmeid.

```python
# Moolimassi arvutamine
# Elementide aatomimassid
masses = {'H': 1.008, 'O': 16.00, 'C': 12.01}
# Koostisosade arv
composition = {'H': 2, 'O': 1}
# Moolimassi arvutamine
molar_mass = sum(masses[element] * count for element, count in composition.items())
molar_mass
```

Moolimass on arvutatud, kasutades elementide aatomimasse ja nende koguseid.

```python
# Tulemuse väljastamine
print(f'Moolimass: {molar_mass} g/mol')
```
