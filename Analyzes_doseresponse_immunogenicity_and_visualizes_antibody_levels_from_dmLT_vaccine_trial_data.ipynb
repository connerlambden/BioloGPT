This notebook downloads the dmLT immunogenicity data and constructs dose-response graphs based on antibody titers.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Download dataset from the provided GitHub repository URL
url = 'https://github.com/PMilletich/dmLT_Serology/raw/master/Supplementary_Data_Biomarkers.csv'
data = pd.read_csv(url)

# Plot dose versus antibody titers for IgG and IgA responses
plt.figure(figsize=(8,6))
plt.plot(data['dose'], data['IgG'], marker='o', linestyle='-', color='red', label='IgG')
plt.plot(data['dose'], data['IgA'], marker='s', linestyle='-', color='green', label='IgA')
plt.xlabel('dmLT Dose (µg)')
plt.ylabel('Antibody Titer (Arbitrary Units)')
plt.title('Dose Response of Antibody Responses in dmLT Trial')
plt.legend()
plt.show()
```

The above code plots the relationship between the administered dmLT dose and the resulting antibody responses to help evaluate the immunogenicity profile.
