Cette analyse explore les points de fusion de divers matériaux en fonction de leurs propriétés thermiques.

```python
import pandas as pd

# Charger les données expérimentales
data = pd.read_csv('melting_points.csv')

# Analyser les points de fusion en fonction de la pression et de la température
results = data.groupby(['Material', 'Pressure']).mean()
```

Les résultats montrent comment les points de fusion varient avec la pression pour différents matériaux.

```python
results.to_csv('melting_analysis.csv')
```
