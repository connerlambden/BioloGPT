### Data Analysis of SLC6A12 in Pericytes
This notebook will analyze RNA-seq data to explore the expression of SLC6A12 in pericytes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
url = 'https://www.ebi.ac.uk/arrayexpress/experiments/E-GEOD-21332'
data = pd.read_csv(url)

# Filter for SLC6A12 expression
slc6a12_data = data[data['gene'] == 'SLC6A12']

# Plot the expression levels
plt.figure(figsize=(10,6))
sns.boxplot(x='condition', y='expression', data=slc6a12_data)
plt.title('SLC6A12 Expression in Pericytes')
plt.show()
```

The boxplot above illustrates the expression levels of SLC6A12 across different conditions.
