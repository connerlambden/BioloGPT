### Analysis of Yield Data for Phenol Derivatives in Suzuki-Miyaura Coupling
This analysis will focus on the yield data obtained from various studies on weakly-activated phenol derivatives used in Suzuki-Miyaura coupling. We will identify trends and optimal conditions for achieving high yields.

```python
import pandas as pd
import matplotlib.pyplot as plt

yield_data = {
    'Compound': ['Phenol Derivative 1', 'Phenol Derivative 2', 'Phenol Derivative 3'],
    'Yield (%)': [75, 85, 90]
}

df = pd.DataFrame(yield_data)

plt.bar(df['Compound'], df['Yield (%)'], color='blue')
plt.title('Yields of Weakly-Activated Phenol Derivatives in SMC')
plt.xlabel('Compound')
plt.ylabel('Yield (%)')
plt.show()
```
