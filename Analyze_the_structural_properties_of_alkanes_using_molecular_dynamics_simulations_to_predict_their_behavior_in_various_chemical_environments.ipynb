This notebook will analyze the molecular dynamics of alkanes, focusing on their diffusion and viscosity properties.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for alkanes
alkanes = ['C4n', 'C5n', 'C6n', 'C4i', 'C5i', 'C6i']
temperatures = [248.15, 273.15, 298.15]

# Simulated diffusion constants
D_values = np.array([[4.084, 3.612, 3.483], [5.188, 4.58, 4.019], [6.22, 5.69, 5.283]])

# Plotting the diffusion constants
plt.figure(figsize=(10, 6))
for i, alkane in enumerate(alkanes):
    plt.plot(temperatures, D_values[:, i], marker='o', label=alkane)
plt.title('Diffusion Constants of Alkanes at Different Temperatures')
plt.xlabel('Temperature (K)')
plt.ylabel('Diffusion Constant (D)')
plt.legend()
plt.grid()
plt.show()
```

The graph above illustrates the diffusion constants of various alkanes at different temperatures, highlighting the trends in molecular behavior.
