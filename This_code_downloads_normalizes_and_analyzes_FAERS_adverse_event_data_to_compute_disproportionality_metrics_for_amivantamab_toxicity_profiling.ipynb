Load and preprocess the FAERS data and filter for amivantamab-related reports.

```python
import pandas as pd
# Download and load data (assuming API endpoint or local file)
df = pd.read_csv('faers_amivantamab.csv')
# Filter for NSCLC and compute metrics such as ROR
filtered = df[df['drug'] == 'amivantamab']
# Compute reporting odds ratio and print summary
print(filtered.groupby('adverse_event').agg({'cases':'sum'}))
```

Visualize the key adverse event frequencies.

```python
import plotly.express as px
fig = px.bar(filtered, x='adverse_event', y='cases', title='Amivantamab Adverse Event Frequencies')
fig.show()
```
