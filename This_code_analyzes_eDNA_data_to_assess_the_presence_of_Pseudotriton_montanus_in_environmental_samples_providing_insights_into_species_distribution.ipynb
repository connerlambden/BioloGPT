### eDNA Analysis for Pseudotriton montanus

```python
import pandas as pd

data = {'Location': ['Kentucky', 'Ohio', 'Tennessee'], 'eDNA_Detected': [1, 0, 1]}
df = pd.DataFrame(data)

# Display the DataFrame
df
```

This DataFrame summarizes the detection of eDNA across different locations.

```python
# Plotting the results
import matplotlib.pyplot as plt

plt.bar(df['Location'], df['eDNA_Detected'], color=['green' if x else 'red' for x in df['eDNA_Detected']])
plt.title('eDNA Detection of Pseudotriton montanus')
plt.xlabel('Location')
plt.ylabel('eDNA Detected (1 = Yes, 0 = No)')
plt.show()
```
