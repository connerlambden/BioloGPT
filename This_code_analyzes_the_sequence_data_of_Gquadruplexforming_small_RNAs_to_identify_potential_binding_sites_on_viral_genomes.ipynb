### G-Quadruplex Sequence Analysis
This notebook will analyze sequences of G-quadruplex-forming small RNAs and their potential interactions with viral genomes.

```python
import pandas as pd

# Load G-quadruplex sequences
g_quadruplex_sequences = pd.read_csv('g_quadruplex_sequences.csv')

# Analyze sequences for potential binding sites
binding_sites = []
for seq in g_quadruplex_sequences['sequence']:
    if 'G' in seq:
        binding_sites.append(seq)

# Display potential binding sites
binding_sites_df = pd.DataFrame(binding_sites, columns=['Potential Binding Sites'])
binding_sites_df
```

### Results
The analysis identifies sequences that may interact with viral RNA, providing insights for further experimental validation.

```python
binding_sites_df.to_csv('potential_binding_sites.csv', index=False)
```
