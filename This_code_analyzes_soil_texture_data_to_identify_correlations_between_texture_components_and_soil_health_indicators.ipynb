This notebook will analyze soil texture data, focusing on the relationships between sand, silt, clay, and organic matter.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load soil texture dataset
data = pd.read_csv('soil_texture_data.csv')

# Analyze correlations
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix of Soil Texture Components')
plt.show()
```

The analysis reveals how different soil texture components interact and their impact on soil health.
