This notebook downloads the 16S rRNA datasets from specified repositories, preprocesses them for quality, and computes diversity indices to compare microbiomes between CFB KO and WT mice.

```python
import qiime2
import pandas as pd
# Download and import dataset
# Perform quality filtering, operational taxonomic unit (OTU) picking, and compute diversity indices.
df = pd.read_csv('path/to/16S_data.csv')
diversity_indices = df.groupby('sample_id').agg({'alpha_diversity':'mean','beta_diversity':'mean'})
print(diversity_indices.head())
```

This section details preprocessing steps and statistical analyses employed to compare the gut and vaginal microbiome diversities.

```python
# Additional analysis code
import scipy.stats as stats
alpha_wt = diversity_indices.loc[df['group']=='WT', 'alpha_diversity']
alpha_ko = diversity_indices.loc[df['group']=='KO', 'alpha_diversity']
stat, p_val = stats.ttest_ind(alpha_wt, alpha_ko)
print('p-value:', p_val)
```

The code produces statistical comparisons to determine significant differences in diversity metrics between groups.

```python
# Visualization using Plotly
import plotly.express as px
fig = px.box(diversity_indices, x='group', y='alpha_diversity', title='Alpha Diversity Comparison')
fig.show()
```
