### HRV Analysis in TBI Patients
This notebook analyzes heart rate variability (HRV) data to explore its relationship with multiple organ dysfunction in TBI patients.

```python
import pandas as pd
import numpy as np

# Load HRV dataset
dataset = pd.read_csv('hrv_data.csv')

# Analyze correlations with organ dysfunction outcomes
correlation_matrix = dataset.corr()
```

### Correlation Analysis
The correlation matrix provides insights into how HRV metrics relate to various organ dysfunction indicators.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix of HRV and Organ Dysfunction')
plt.show()
```
