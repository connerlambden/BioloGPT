This section describes steps to download RNA-seq data from BioProject PRJNA1124505, process the high-depth reads, and perform alternative splicing analysis.

```python
import subprocess

# Download dataset using fastq-dump
subprocess.run(['fastq-dump', '--split-files', 'SRRXXXXXX'])

# Run Salmon for transcript quantification
subprocess.run(['salmon', 'quant', '-i', 'transcript_index', '-l', 'A', '-r', 'reads.fq', '-o', 'quant_output'])
```

Next, the notebook demonstrates differential splicing analysis using the 3D RNA-seq toolkit and visualization of isoform usage differences between WT and mutant samples.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load processed data
df = pd.read_csv('differential_splicing_results.csv')

# Plot isoform usage differences
plt.bar(df['Isoform'], df['Usage'], color=['#6A0C76', '#B565A7'])
plt.xlabel('DGK5 Isoforms')
plt.ylabel('Relative Usage (%)')
plt.title('Isoform Usage in WT vs cpl3 Mutant')
plt.show()
```
