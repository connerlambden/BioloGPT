The following code downloads and processes dataset information on neuronal excitability changes following dopamine receptor activation.

```python
import pandas as pd
# Download dataset from provided study link
url = 'https://biologpt.com/data/ceA_dopamine_excitability.csv'
df = pd.read_csv(url)
print(df.head())
```

We then analyze differences in input resistance, rheobase, and EPSC parameters between D1 and D2 activation conditions.

```python
import matplotlib.pyplot as plt
# Plot input resistance changes
plt.figure(figsize=(6,4))
plt.boxplot([df[df['agonist']=='D1']['input_resistance'], df[df['agonist']=='D2']['input_resistance']], labels=['D1','D2'])
plt.title('Input Resistance under D1 vs D2 Activation')
plt.ylabel('Input Resistance (MΩ)')
plt.show()
```

This analysis helps visualize differential impacts of receptor subtype activation on neuronal excitability.
