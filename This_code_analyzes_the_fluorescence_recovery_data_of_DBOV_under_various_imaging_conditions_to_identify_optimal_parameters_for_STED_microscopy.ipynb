This notebook analyzes fluorescence recovery data from DBOV under different reactivation conditions.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
fluorescence_data = pd.read_csv('dbov_fluorescence_data.csv')

# Plot fluorescence recovery
plt.plot(fluorescence_data['time'], fluorescence_data['intensity'])
plt.title('Fluorescence Recovery of DBOV')
plt.xlabel('Time (s)')
plt.ylabel('Fluorescence Intensity')
plt.show()
```
