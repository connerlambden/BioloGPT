This notebook section downloads and processes survival and developmental data from the study's supplementary files, generating dose-response graphs for each compound.

```python
import pandas as pd
import plotly.express as px

data = pd.read_csv('https://www.mdpi.com/article/10.3390/jox15010033/s1/dose_response.csv')
fig = px.line(data, x='Concentration', y='Survival', color='Compound', title='Dose-Response Curve for Synthetic Cathinones on C. elegans')
fig.show()
```
