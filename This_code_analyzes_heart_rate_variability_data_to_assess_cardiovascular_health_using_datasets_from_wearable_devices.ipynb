This notebook will analyze heart rate data collected from wearable devices to evaluate cardiovascular health.

```python
import pandas as pd

# Load heart rate data
data = pd.read_csv('heart_rate_data.csv')

# Analyze heart rate variability
hrv = data['heart_rate'].std()  # Standard deviation as a measure of variability
print(f'Heart Rate Variability: {hrv}')
```

The analysis provides insights into the variability of heart rates, which is crucial for understanding cardiovascular health.
