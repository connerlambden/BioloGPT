```python
import pandas as pd
import plotly.express as px

data = pd.DataFrame({
    'Site': ['Kapsabet (KP1A)', 'Homa Bay (HK2)', 'Siaya (KK1)', 'Busia (SPK1)', 'Kisumu (DK2)', 'Lake Victoria (LV95-2P)'],
    'PAHs': [270, 283, 200, 60, 31, 353],
    'nAlkanes': [51, 258, 40, 5, 420, 168]
})

fig = px.bar(data, x='Site', y=['PAHs','nAlkanes'], barmode='group', title='Chemical Marker Trends')
fig.show()
```

This notebook helps visualize how PAH and n-alkane variations correspond with different sites, aiding hypothesis testing regarding urban versus rural impacts.

```python
# Further statistical correlation analysis can be added here using scipy or statsmodels
import scipy.stats as stats
correlation, p_value = stats.pearsonr(data['PAHs'], data['nAlkanes'])
print('Correlation between PAHs and n-Alkanes:', correlation)
print('P-value:', p_value)
```
