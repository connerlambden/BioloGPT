### Longevity Substances Analysis
This notebook analyzes the effects of various substances on lifespan based on existing research.

```python
import pandas as pd

# Sample data for analysis
data = {
    'Substance': ['3-Phenyllactic Acid', 'Selegiline', 'Metformin', 'Resveratrol', 'Tanespimycin'],
    'Lifespan Extension (Weeks)': [23, 14, 12, 10, 8]
}
df = pd.DataFrame(data)

# Display the dataframe
df
```

### Summary of Findings
The analysis shows that 3-Phenyllactic Acid has the highest reported lifespan extension.

```python
# Plotting the data
import matplotlib.pyplot as plt

plt.bar(df['Substance'], df['Lifespan Extension (Weeks)'])
plt.title('Impact of Substances on Longevity')
plt.xlabel('Substances')
plt.ylabel('Lifespan Extension (Weeks)')
plt.xticks(rotation=45)
plt.show()
```
