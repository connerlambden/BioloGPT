This notebook will explore the impact of ApoE4 on cognitive decline using data from the Alzheimer's Disease Neuroimaging Initiative.

```python
import pandas as pd
# Load relevant datasets
adni_data = pd.read_csv('adni_data.csv')
```

We will analyze the cognitive scores of ApoE4 carriers versus non-carriers.

```python
cognitive_scores = adni_data.groupby('ApoE4_status')['cognitive_score'].mean()
```
