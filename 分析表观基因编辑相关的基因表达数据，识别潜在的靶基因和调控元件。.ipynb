使用RNA-seq数据分析表观基因编辑对基因表达的影响。

```python
import pandas as pd
# 读取RNA-seq数据
data = pd.read_csv('rna_seq_data.csv')
# 进行数据分析
results = data.groupby('gene').mean()
```

分析结果显示哪些基因在表观基因编辑后表达显著变化。

```python
results.to_csv('analysis_results.csv')
```
