### Analysis of Plasmid Purification Methods
This notebook will analyze the yield and purity of plasmid DNA obtained from various purification methods, using data extracted from recent studies.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for plasmid purification methods
data = {
    'Method': ['Alkaline Lysis', 'Miniprep Kit', 'Continuous Extraction'],
    'Yield (%)': [80, 90, 85],
    'Purity (A260/A280)': [1.8, 2.0, 1.9]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting the data
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Method')
ax1.set_ylabel('Yield (%)', color=color)
ax1.bar(df['Method'], df['Yield (%)'], color=color, alpha=0.6, label='Yield')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  
color = 'tab:red'
ax2.set_ylabel('Purity (A260/A280)', color=color)  
ax2.plot(df['Method'], df['Purity (A260/A280)'], color=color, marker='o', label='Purity')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  
plt.title('Comparison of Plasmid Purification Methods')
plt.show()
```
