### RNA-seq Data Analysis for THAP9 and Myelin Genes

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load RNA-seq data
thap9_data = pd.read_csv('thap9_expression.csv')
myelin_data = pd.read_csv('myelin_genes_expression.csv')

# Merge datasets
merged_data = pd.merge(thap9_data, myelin_data, on='sample_id')

# Plot expression levels
plt.figure(figsize=(10,6))
sns.boxplot(data=merged_data, x='cell_type', y='expression_level', hue='gene')
plt.title('Expression Levels of THAP9 and Myelin Genes')
plt.show()
```

This analysis visualizes the expression levels of THAP9 and myelin-associated genes across different cell types, providing insights into their regulatory relationships.

```python
# Save the plot
plt.savefig('thap9_myelin_expression.png')
```
