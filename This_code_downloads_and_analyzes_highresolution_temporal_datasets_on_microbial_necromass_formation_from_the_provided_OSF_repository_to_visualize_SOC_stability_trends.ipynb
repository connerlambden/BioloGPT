Import datasets on microbial necromass and SOC stability, then preprocess and analyze temporal changes using statistical models.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Download dataset (assuming a URL from OSF repository provided in the paper)
data_url = 'https://osf.io/XXX'
data = pd.read_csv(data_url)
# Process data
summary = data.groupby('treatment')['SOC_stability'].mean()
summary.plot(kind='bar', color='#6A0C76', title='SOC Stability by Treatment')
plt.xlabel('Treatment')
plt.ylabel('Mean SOC Stability')
plt.show()
```

This analysis helps visualize the impact of microbial community composition on SOC resistance to decomposition.

```python
# Further statistical analysis
import statsmodels.api as sm
X = data[['moisture', 'microbial_diversity']]
X = sm.add_constant(X)
model = sm.OLS(data['SOC_stability'], X).fit()
print(model.summary())
```
