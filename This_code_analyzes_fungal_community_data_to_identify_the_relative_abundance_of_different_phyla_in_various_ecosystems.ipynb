This notebook analyzes fungal community data from various studies to visualize the relative abundance of Ascomycota, Basidiomycota, Zygomycota, and Chytridiomycota.

```python
import pandas as pd
import plotly.express as px

# Sample data
data = {'Phylum': ['Ascomycota', 'Basidiomycota', 'Zygomycota', 'Chytridiomycota'],
        'Relative Abundance': [75, 13, 1, 1]}
df = pd.DataFrame(data)

# Create a pie chart
fig = px.pie(df, values='Relative Abundance', names='Phylum', title='Fungal Phyla Relative Abundance')
fig.show()
```
