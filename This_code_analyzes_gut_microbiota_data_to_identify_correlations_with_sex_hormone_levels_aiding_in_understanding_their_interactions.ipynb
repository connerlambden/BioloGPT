This notebook analyzes gut microbiota composition and its correlation with sex hormone levels using 16S rRNA sequencing data.

```python
import pandas as pd
# Load gut microbiota and hormone data
gut_data = pd.read_csv('gut_microbiota.csv')
hormone_data = pd.read_csv('hormone_levels.csv')
# Merge datasets based on sample ID
merged_data = pd.merge(gut_data, hormone_data, on='sample_id')
# Perform correlation analysis
correlation = merged_data.corr()
```

The correlation matrix provides insights into how specific gut bacteria relate to hormone levels.

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Gut Microbiota and Sex Hormones')
plt.show()
```
