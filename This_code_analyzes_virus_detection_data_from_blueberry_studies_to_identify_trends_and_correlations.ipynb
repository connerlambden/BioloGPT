This notebook will analyze virus detection data, focusing on prevalence across different regions and detection methods.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load virus detection data
data = {'Virus': ['BFDaV', 'BlLV', 'BLMoV', 'BlMaV', 'BNRBV', 'BRRV', 'BlScV', 'BlShV', 'BlSSV', 'BVA', 'PRMV', 'TRSV', 'ToRSV'],
        'Positives': [64, 92, 30, 84, 26, 47, 114, 566, 53, 119, 8, 46, 24]}
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10,6))
plt.barh(df['Virus'], df['Positives'], color='skyblue')
plt.xlabel('Number of Positive Samples')
plt.title('Virus Detection in Blueberries')
plt.show()
```
