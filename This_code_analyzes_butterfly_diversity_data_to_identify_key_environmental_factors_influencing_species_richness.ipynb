### Butterfly Diversity Analysis
This notebook analyzes butterfly diversity data to identify key environmental factors influencing species richness.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load butterfly diversity dataset
data = pd.read_csv('butterfly_diversity.csv')

# Analyze species richness by habitat type
sns.boxplot(x='Habitat', y='Species_Richness', data=data)
plt.title('Butterfly Species Richness by Habitat Type')
plt.show()
```

### Discussion
This analysis provides insights into how different habitats support butterfly diversity.
