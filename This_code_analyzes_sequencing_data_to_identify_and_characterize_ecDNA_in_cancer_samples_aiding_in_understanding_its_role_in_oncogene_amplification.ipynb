### Data Preparation
Load the sequencing data and preprocess it for analysis.

```python
import pandas as pd
# Load sequencing data
data = pd.read_csv('ecDNA_sequencing_data.csv')
```

### Analysis of ecDNA Presence
Identify ecDNA sequences and their associated oncogenes.

```python
ecDNA_sequences = data[data['type'] == 'ecDNA']
# Count occurrences of oncogenes
oncogene_counts = ecDNA_sequences['oncogene'].value_counts()
```

### Visualization
Visualize the distribution of ecDNA across samples.

```python
import matplotlib.pyplot as plt
plt.bar(oncogene_counts.index, oncogene_counts.values)
plt.title('Distribution of Oncogenes on ecDNA')
plt.xlabel('Oncogenes')
plt.ylabel('Count')
plt.show()
```
