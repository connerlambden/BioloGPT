## Analysis of Nrf2 Regulatory Networks in Astrocytes
This notebook explores the gene expression changes in Nrf2 knockout astrocytes to identify key regulatory pathways.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the extracted data
data = {
    'Genotype': ['WT', 'HET', 'KO'],
    'Intracellular_GSH': [0.8673, 0.0, 0.5],  # Example value for KO
    'Extracellular_GSH': [0.9971, 0.0250, 0.2856],
    'xCT_mRNA': [1.0, 0.5, 0.2],
    'GCLc_mRNA': [1.0, 1.0, 0.3],
    'GCLm_mRNA': [1.0, 0.3, 0.2]
}

df = pd.DataFrame(data)
```

### Visualizing GSH Levels

```python
sns.barplot(x='Genotype', y='Intracellular_GSH', data=df, color='skyblue')
plt.title('Intracellular GSH Levels in Astrocytes')
plt.ylabel('GSH (µmol/g protein)')
plt.show()
```

### Correlation Between Gene Expression and GSH Levels

```python
genes = ['xCT_mRNA', 'GCLc_mRNA', 'GCLm_mRNA']
corr = df[genes + ['Intracellular_GSH']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

### Conclusion
The analysis highlights significant correlations between the expression levels of <em>xCT</em>, <em>GCLc</em>, and <em>GCLm</em> with intracellular GSH levels, underscoring the role of Nrf2 in regulating these genes to maintain glutathione homeostasis in astrocytes.
