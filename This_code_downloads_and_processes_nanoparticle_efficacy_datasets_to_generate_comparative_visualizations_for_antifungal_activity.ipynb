This notebook extracts data from the provided nanoparticle studies and organizes it into visual tables and graphs.

```python
import pandas as pd
import plotly.express as px

# Load dataset
data = pd.DataFrame([
    {'Treatment': 'Control', 'Incubation': 0, 'Latent': 0, 'Pustule Size': 100},
    {'Treatment': 'Ag NP', 'Incubation': 2, 'Latent': 3, 'Pustule Size': 70},
    {'Treatment': 'Cu NP', 'Incubation': 2, 'Latent': 3, 'Pustule Size': 70},
    {'Treatment': 'Combined', 'Incubation': 4, 'Latent': 5, 'Pustule Size': 40}
])

# Plot graph
fig = px.bar(data, x='Treatment', y='Incubation', title='Impact of Nanoparticles on Incubation Period')
fig.show()
```

The code above provides a basic visualization of how nanoparticle treatments affect the incubation period relative to a control.
