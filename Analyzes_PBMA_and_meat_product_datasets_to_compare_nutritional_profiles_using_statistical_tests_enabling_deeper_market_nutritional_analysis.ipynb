This notebook downloads and processes the PBMA and conventional meat datasets to compute and compare key nutritional metrics using t-tests.

```python
import pandas as pd
from scipy.stats import ttest_ind

# Load datasets
pbma_data = pd.read_csv('pbma_products.csv')
meat_data = pd.read_csv('meat_products.csv')

# Perform t-test on saturated fat
t_stat, p_value = ttest_ind(pbma_data['saturated_fat'], meat_data['saturated_fat'])
print('T-test statistic:', t_stat, 'P-value:', p_value)

# Further analysis can be added here to compare other nutritional metrics
```
