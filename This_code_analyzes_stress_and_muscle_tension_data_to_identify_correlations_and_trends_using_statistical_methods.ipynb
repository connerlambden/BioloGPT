### Data Analysis of Stress and Muscle Tension
This notebook will analyze the correlation between stress levels and facial muscle tension using collected data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('stress_muscle_tension.csv')

# Analyze the correlation
correlation = data.corr()

# Visualize the correlation matrix
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Stress Levels and Facial Muscle Tension')
plt.show()
```

### Results Interpretation
The heatmap will provide insights into how stress levels correlate with facial muscle tension, guiding future interventions.
