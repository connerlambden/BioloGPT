### Data Analysis of H5N1 Transmission and Wind Conditions
This notebook analyzes the relationship between wind speed, direction, and H5N1 transmission events using environmental datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
wind_data = pd.read_csv('wind_data.csv')
h5n1_data = pd.read_csv('h5n1_cases.csv')

# Merge datasets on date and location
merged_data = pd.merge(wind_data, h5n1_data, on=['date', 'location'])

# Plotting the data
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['wind_speed'], merged_data['cases'], alpha=0.5)
plt.title('H5N1 Cases vs Wind Speed')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Number of H5N1 Cases')
plt.grid()
plt.show()
```
