### Protease Activity Analysis
This notebook analyzes protease activity data from various studies to identify trends and correlations.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Strain': ['Wild Type', 'Mutant rsepA1'], 'Activity': [14.86, 117.0]}
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Strain'], df['Activity'], color=['blue', 'orange'])
plt.title('Protease Activity Comparison')
plt.ylabel('Activity (IU/ml)')
plt.show()
```

### Discussion
This analysis highlights the significant increase in protease activity in the mutant strain, suggesting potential applications in biotechnology.
