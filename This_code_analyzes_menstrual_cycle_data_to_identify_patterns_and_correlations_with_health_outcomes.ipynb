### Menstrual Cycle Data Analysis
This notebook analyzes menstrual cycle data to explore correlations with health outcomes.

```python
import pandas as pd

# Load menstrual cycle data
data = pd.read_csv('menstrual_cycle_data.csv')

# Analyze cycle length and symptoms
cycle_length = data['cycle_length']
symptoms = data['symptoms']

# Correlation analysis
correlation = cycle_length.corr(symptoms)
```
