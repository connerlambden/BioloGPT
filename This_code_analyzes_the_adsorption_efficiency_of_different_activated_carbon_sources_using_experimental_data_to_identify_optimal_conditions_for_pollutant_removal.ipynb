### Data Analysis of Activated Carbon Adsorption Efficiency
This notebook will analyze the adsorption efficiency of various activated carbon sources based on experimental data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset containing adsorption efficiency data
data = pd.read_csv('adsorption_efficiency.csv')

# Analyze the data
summary = data.describe()

# Plotting the results
plt.figure(figsize=(10, 6))
plt.bar(data['source'], data['efficiency'])
plt.title('Adsorption Efficiency of Different Activated Carbon Sources')
plt.xlabel('Source of Activated Carbon')
plt.ylabel('Efficiency (%)')
plt.xticks(rotation=45)
plt.show()
```
