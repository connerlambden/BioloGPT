### Metabolite-Disease Association Analysis
This notebook analyzes the associations between metabolites and diseases using data from HMDB and DisGeNET.

```python
import pandas as pd

# Load metabolite data from HMDB
hmdb_data = pd.read_csv('hmdb_data.csv')

# Load disease association data from DisGeNET
disgenet_data = pd.read_csv('disgenet_data.csv')

# Merge datasets on common metabolites
merged_data = pd.merge(hmdb_data, disgenet_data, on='metabolite_id')

# Analyze associations
associations = merged_data[['metabolite_name', 'disease_name', 'association_strength']]
associations.head()
```

### Results
The analysis provides a list of metabolites associated with various diseases, highlighting potential biomarkers.

```python
# Display results
associations.to_csv('metabolite_disease_associations.csv', index=False)
```
