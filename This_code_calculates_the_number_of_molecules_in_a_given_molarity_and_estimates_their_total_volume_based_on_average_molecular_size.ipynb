### Calculate Number of Molecules and Volume
This notebook calculates the number of molecules in a solution based on molarity and estimates the total volume occupied by these molecules.

```python
# Constants
molarity = 5e-11  # M
avogadro_number = 6.022e23  # molecules/mole
molecular_size_nm = 5  # average size in nm
radius_nm = molecular_size_nm / 2

# Calculate number of molecules
number_of_molecules = molarity * avogadro_number

# Volume of one molecule (sphere)
import math
volume_one_molecule = (4/3) * math.pi * (radius_nm ** 3)

# Total volume occupied by all molecules
total_volume = number_of_molecules * volume_one_molecule

number_of_molecules, total_volume
```
