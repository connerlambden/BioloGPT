This notebook will explore the antibacterial properties of various bioactive compounds found in lupine seeds, utilizing datasets from recent studies.

```python
import pandas as pd
# Load relevant datasets
lupine_data = pd.read_csv('lupine_bioactive_compounds.csv')
# Analyze antibacterial activity
results = lupine_data.groupby('compound').agg({'antibacterial_activity': 'mean'})
```

The analysis will provide insights into which compounds exhibit the strongest antibacterial effects and their potential mechanisms.

```python
results.to_csv('lupine_antibacterial_analysis.csv')
# Save the results for further analysis.
```
