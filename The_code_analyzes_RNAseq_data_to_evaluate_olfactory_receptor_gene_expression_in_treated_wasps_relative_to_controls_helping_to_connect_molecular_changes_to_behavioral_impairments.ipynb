This notebook downloads RNA-seq datasets related to olfactory receptor expression in Leptopilina heterotoma exposed to insecticides.

```python
import pandas as pd
# Code to download dataset from a given URL (replace with actual URL)
df = pd.read_csv('https://example.com/olfactory_receptors.csv')
df.head()
```

Next, perform differential expression analysis to compare treated vs control groups.

```python
import statsmodels.api as sm
# Assume df has columns 'expression', 'condition'
model = sm.OLS(df['expression'], sm.add_constant(pd.get_dummies(df['condition'], drop_first=True)))
results = model.fit()
print(results.summary())
```

This analysis helps link gene expression changes to observed behavioral deficits.

```python
# Visualization using plotly
import plotly.express as px
fig = px.bar(df, x='gene', y='expression', color='condition', barmode='group')
fig.show()
```
