### Morphological Data Analysis of Mallomonas asmundiae
This notebook analyzes the scale measurements of fossil and modern specimens.

```python
import pandas as pd

# Data for analysis
fossil_data = {'Scale Length (μm)': [5.3], 'Scale Width (μm)': [2.7]}
modern_data = {'Scale Length (μm)': [5.0], 'Scale Width (μm)': [2.5]}

# Create DataFrames
fossil_df = pd.DataFrame(fossil_data)
modern_df = pd.DataFrame(modern_data)

# Combine DataFrames
combined_df = pd.concat([fossil_df, modern_df], keys=['Fossil', 'Modern'])
combined_df
```
