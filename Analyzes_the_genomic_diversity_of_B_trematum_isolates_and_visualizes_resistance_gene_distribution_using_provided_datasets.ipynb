## Genomic Diversity Analysis of B. trematum Isolates

```python
import pandas as pd
import plotly.express as px

data = [
    {'isolate': 'PIW211', 'genome_length': 4350381, 'GC_content': 65.45, 'contigs': 64, 'N50': 193049, 'coding_sequences': 4036},
    {'isolate': 'PIW212', 'genome_length': 4311189, 'GC_content': 65.49, 'contigs': 67, 'N50': 278393, 'coding_sequences': 3984}
]

df = pd.DataFrame(data)
```

### GC Content Comparison

```python
fig = px.bar(df, x='isolate', y='GC_content', title='GC Content of B. trematum Isolates', labels={'GC_content': 'GC Content (%)'})
fig.show()
```

### Genome Assembly Metrics

```python
fig = px.scatter(df, x='genome_length', y='N50', size='contigs', color='isolate', title='Genome Length vs N50', labels={'genome_length': 'Genome Length (bp)', 'N50': 'N50 (bp)'})
fig.show()
```
