This notebook will analyze RNA-seq data from SLE patients to explore the expression of genes related to Treg and Th17 differentiation.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('SLE_gene_expression.csv')

# Analyze the expression of key genes
key_genes = ['FOXP3', 'RORC', 'IL17A']
expression_data = data[data['gene'].isin(key_genes)]

# Visualize the expression levels
sns.boxplot(x='gene', y='expression', data=expression_data)
plt.title('Gene Expression in SLE Patients')
plt.show()
```

This analysis will help identify potential biomarkers for Treg and Th17 cell activity in SLE patients.
