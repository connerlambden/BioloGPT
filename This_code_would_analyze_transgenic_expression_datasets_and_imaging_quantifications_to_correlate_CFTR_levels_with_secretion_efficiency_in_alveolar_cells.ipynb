Import imaging quantification data and gene expression profiles from transgenic mice to correlate CFTR function with liquid secretion rates.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('imaging_quantification.csv')
expression = pd.read_csv('AT1_CFTR_expression.csv')
merged = pd.merge(data, expression, on='sample_id')
plt.scatter(merged['CFTR_expression'], merged['secretion_rate'])
plt.xlabel('CFTR Expression')
plt.ylabel('Liquid Secretion Rate')
plt.title('Correlation between CFTR Expression and Secretion Rate in AT1 Cells')
plt.show()
```

This notebook code visualizes the correlation between CFTR expression levels in AT1 cells and measured liquid secretion rates.
