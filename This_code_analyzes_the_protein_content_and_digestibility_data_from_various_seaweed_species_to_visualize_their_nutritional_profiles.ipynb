### Analysis of Seaweed Protein Content and Digestibility
This notebook will analyze the protein content and digestibility of different seaweed species and their derived mycoproteins. The data will be visualized to compare their nutritional profiles.

```python
import pandas as pd
import plotly.express as px

# Data
data = {
    'Sample': ['Seaweed-derived mycoprotein', 'Ulva spp.', 'Macrocystis pyrifera', 'Durvillaea spp.'],
    'Crude Protein (g/100g dw)': [31.35, 20.9, 12, 11.8],
    'PDCAAS': [0.89, 0.37, 0.41, 0.53]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot
fig = px.bar(df, x='Sample', y=['Crude Protein (g/100g dw)', 'PDCAAS'], barmode='group', title='Protein Content and PDCAAS of Seaweeds and Mycoprotein')
fig.show()
```
