### Fragrance Compound Analysis
This notebook analyzes the properties of various fragrance compounds to identify trends.

```python
import pandas as pd

# Sample data of fragrance compounds
data = {
    'Compound': ['Patchouli Oil', 'Rose Oil', 'Geraniol', 'Citronellol'],
    'Source': ['Natural', 'Natural', 'Synthetic', 'Natural'],
    'Popularity': [30, 25, 20, 15]
}
df = pd.DataFrame(data)

# Analyzing the data
print(df.groupby('Source').sum())
```

### Analysis Results
The analysis shows the popularity of natural vs synthetic compounds.

```python
# Displaying the results
import matplotlib.pyplot as plt

plt.bar(df['Compound'], df['Popularity'])
plt.title('Popularity of Fragrance Compounds')
plt.xlabel('Compound')
plt.ylabel('Popularity')
plt.show()
```
