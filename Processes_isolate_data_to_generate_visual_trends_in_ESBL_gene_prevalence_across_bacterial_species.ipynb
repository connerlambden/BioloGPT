Load bacterial isolate data and visualize resistance gene distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define dataset
data = {
    'Organism': ['Bacillus sp', 'Alcaligenes sp', 'Proteus spp', 'Providencia spp', 'Enterococcus faecalis', 'Unidentified', 'Lost sample'],
    'Count': [24, 23, 2, 2, 2, 3, 3]
}
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(8,4))
plt.bar(df['Organism'], df['Count'], color='#6A0C76')
plt.xlabel('Organism')
plt.ylabel('Count')
plt.title('Bacterial Isolate Counts from Gastroenteritis Study')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
