Analyze P-selectin expression data from various studies to identify correlations with thrombotic events.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('p_selectin_expression.csv')
sns.boxplot(x='condition', y='P_selectin_level', data=data)
plt.title('P-selectin Expression Levels by Condition')
plt.show()
```
