This notebook imports mutation data for TCOF1, processes mutation details (e.g., location, type) and generates plots to visualize mutation distribution across protein domains.

```python
import pandas as pd
import plotly.express as px
# Load mutation dataset
data = pd.DataFrame([
    {'Patient': 'TCS 1', 'Domain': 'CRD', 'CADD': 23.4},
    {'Patient': 'TCS 2', 'Domain': 'CRD', 'CADD': 22.4},
    {'Patient': 'TCS 3', 'Domain': 'CRD', 'CADD': 20.9},
    {'Patient': 'TCS 5', 'Domain': 'CRD', 'CADD': 26.1},
    {'Patient': 'TCS 6', 'Domain': 'CRD', 'CADD': None},
    {'Patient': 'TCS 8', 'Domain': 'CRD', 'CADD': None},
    {'Patient': 'TCS 9', 'Domain': '-', 'CADD': None},
    {'Patient': 'TCS 10', 'Domain': 'CTD (NLS)', 'CADD': None},
    {'Patient': 'TCS 11', 'Domain': 'CTD (NLS)', 'CADD': None},
    {'Patient': 'TCS 12', 'Domain': 'CTD (NLS)', 'CADD': 22.5}
])
fig = px.bar(data, x='Patient', y='CADD', color='Domain', title='TCOF1 Mutation CADD Scores by Domain')
fig.show()
```
