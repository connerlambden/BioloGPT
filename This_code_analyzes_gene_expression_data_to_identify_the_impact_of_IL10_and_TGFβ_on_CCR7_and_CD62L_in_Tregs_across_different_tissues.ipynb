This notebook will analyze gene expression datasets to explore how IL-10 and TGF-β influence CCR7 and CD62L expression in Tregs.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
il10_data = pd.read_csv('il10_expression.csv')
tgf_beta_data = pd.read_csv('tgf_beta_expression.csv')

# Analyze expression levels
merged_data = pd.merge(il10_data, tgf_beta_data, on='Gene')
sns.boxplot(x='Tissue', y='Expression', hue='Cytokine', data=merged_data)
plt.title('Cytokine Influence on CCR7 and CD62L Expression in Tregs')
plt.show()
```
