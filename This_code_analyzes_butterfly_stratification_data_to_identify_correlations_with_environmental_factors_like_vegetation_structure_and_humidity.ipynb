### Butterfly Stratification Analysis
This notebook analyzes the relationship between butterfly stratification and various environmental factors.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load butterfly data
butterfly_data = pd.read_csv('butterfly_stratification_data.csv')

# Analyze correlations
correlation_matrix = butterfly_data.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix of Environmental Factors and Butterfly Stratification')
plt.show()
```
