### Data Analysis for Electroporation Efficiency
This section will analyze transformation efficiency data to determine the best electroporation conditions.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Voltage (kV)': [2.5, 2.5, 2.0], 'Transformation Efficiency (CFU/µg)': [1e5, 5e4, 2e5]}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.bar(df['Voltage (kV)'], df['Transformation Efficiency (CFU/µg)'], color='purple')
plt.xlabel('Voltage (kV)')
plt.ylabel('Transformation Efficiency (CFU/µg)')
plt.title('Transformation Efficiency vs Voltage')
plt.show()
```
