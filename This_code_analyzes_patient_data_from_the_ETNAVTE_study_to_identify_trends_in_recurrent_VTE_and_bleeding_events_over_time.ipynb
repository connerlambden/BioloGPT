### ETNA-VTE Data Analysis
This notebook analyzes the data from the ETNA-VTE study to visualize trends in recurrent VTE and major bleeding events.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data from ETNA-VTE study
data = {'Event': ['Recurrent VTE', 'Major Bleeding'], 'Annualized Rate (%)': [2.7, 1.5]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Event'], df['Annualized Rate (%)'], color=['blue', 'orange'])
plt.title('Annualized Rates of Recurrent VTE and Major Bleeding')
plt.ylabel('Annualized Rate (%)')
plt.show()
```
