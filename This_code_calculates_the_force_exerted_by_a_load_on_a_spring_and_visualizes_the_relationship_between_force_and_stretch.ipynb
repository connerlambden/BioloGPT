# Spring Force Calculation
This notebook calculates the force exerted by a 20g load on a spring and visualizes the relationship between force and stretch.

```python
# Constants
mass = 0.02  # mass in kg (20g)
g = 9.81  # acceleration due to gravity in m/s^2
force = mass * g  # force in Newtons
force
```

The force exerted by the 20g load is calculated as follows:

```python
# Calculate force
force = mass * g
force
```

The calculated force is approximately 0.1962 N.

```python
# Visualization of Force vs. Stretch
import matplotlib.pyplot as plt
stretch = [0, 0.1, 0.2, 0.3, 0.4]  # stretch in meters
forces = [0, 0.1962, 0.3924, 0.5886, 0.7848]  # corresponding forces
plt.plot(stretch, forces, marker='o')
plt.title('Force vs. Stretch in a Spring')
plt.xlabel('Stretch (m)')
plt.ylabel('Force (N)')
plt.grid()
plt.show()
```
