### Bacterial Viability Analysis
This notebook analyzes the impact of nitrogen oxides and light on bacterial viability using provided datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('bacterial_viability_data.csv')

# Analyze the data
summary = data.describe()

# Plotting the results
plt.figure(figsize=(10, 6))
plt.bar(data['Condition'], data['Viability'])
plt.title('Bacterial Viability under Different Conditions')
plt.xlabel('Condition')
plt.ylabel('Viability (%)')
plt.show()
```

### Summary of Findings
This analysis provides insights into how different environmental conditions affect bacterial viability.

```python
# Save the summary to a file
summary.to_csv('viability_summary.csv')
```
