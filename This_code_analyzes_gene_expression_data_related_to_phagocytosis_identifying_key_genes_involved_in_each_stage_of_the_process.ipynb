This notebook will analyze gene expression data from macrophages during phagocytosis, focusing on genes upregulated during each stage.

```python
import pandas as pd

# Load gene expression data
gene_data = pd.read_csv('gene_expression_phagocytosis.csv')

# Analyze expression levels during different stages of phagocytosis
phagocytosis_stages = gene_data.groupby('stage').mean()
phagocytosis_stages
```

The analysis reveals which genes are most active during the recognition, engulfment, and degradation stages of phagocytosis.

```python
# Visualize the results
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(phagocytosis_stages.index, phagocytosis_stages['expression_level'])
plt.title('Gene Expression Levels During Phagocytosis Stages')
plt.xlabel('Phagocytosis Stage')
plt.ylabel('Average Expression Level')
plt.show()
```
