### Analysis of Painted Turtle Hatchling Survival Data

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Nest': [1, 2, 3, 4, 5, 6, 7],
    'Min_Temperature': [-2.5, -3.0, -3.0, -4.0, -4.0, -6.0, -6.0],
    'Survival_Rate': [90, 67, 0, 100, 50, 29, 40]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.bar(df['Nest'], df['Survival_Rate'], color='skyblue')
plt.title('Survival Rates of Painted Turtle Hatchlings by Nest')
plt.xlabel('Nest')
plt.ylabel('Survival Rate (%)')
plt.xticks(df['Nest'])
plt.grid(axis='y')
plt.show()
```
