The following notebook imports experimental data on NST neuron marker expression and sleep changes to create interactive visualizations.

```python
import plotly.graph_objs as go

# Expression data
markers = ['Slc17a6', 'Slc32a1', 'Adcyap1', 'Cartpt']
expression = [63.3, 36.3, 56.6, 21.7]

fig = go.Figure([go.Bar(x=markers, y=expression, marker_color='#6A0C76')])
fig.update_layout(title='NST LPS-TRAP Neuron Marker Expression', xaxis_title='Markers', yaxis_title='Expression (%)')
fig.show()
```

```python
import plotly.graph_objs as go

sleep_metrics = ['NREM Increase', 'Wake Decrease']
# Simplified change representations; positive values denote increased NREM sleep, negative for wakefulness changes
changes = [1, -1]

fig = go.Figure([go.Bar(x=sleep_metrics, y=changes, marker_color='#6A0C76')])
fig.update_layout(title='Sleep Changes Post NST Activation', xaxis_title='Sleep Metrics', yaxis_title='Direction of Change')
fig.show()
```
