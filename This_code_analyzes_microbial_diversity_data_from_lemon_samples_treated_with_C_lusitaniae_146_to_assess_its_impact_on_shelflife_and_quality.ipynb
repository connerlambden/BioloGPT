### Microbial Diversity Analysis of Lemon Samples
This analysis will utilize sequencing data to compare the microbial communities of treated and untreated lemon samples.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load microbial diversity data
data = pd.read_csv('lemon_microbial_diversity.csv')

# Plotting the diversity indices
plt.figure(figsize=(10,6))
plt.bar(data['Sample'], data['Diversity_Index'], color=['blue', 'orange'])
plt.title('Microbial Diversity Indices in Lemon Samples')
plt.xlabel('Sample Type')
plt.ylabel('Diversity Index')
plt.show()
```
