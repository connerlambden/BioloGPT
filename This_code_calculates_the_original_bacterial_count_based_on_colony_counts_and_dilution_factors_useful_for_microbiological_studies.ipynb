### Bacterial Count Calculation
This notebook calculates the original bacterial count based on the number of colonies counted and the dilution factor.

```python
# Constants
colonies_counted = 86

# Dilution factor
# 10^-7 dilution means 10**7 for calculation
 dilution_factor = 10**7

# Calculate original count
original_count = colonies_counted * dilution_factor
original_count
```

### Result
The original count of bacteria in the undiluted culture is calculated.

```python
# Display the result
print(f'Original Count: {original_count} CFU/ml')
```
