### Analyzing Fungal Diversity on Lemon Surfaces
This notebook will analyze the fungal diversity data collected from lemon samples treated with C. lusitaniae 146.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('lemon_fungal_diversity.csv')

# Analyze the diversity metrics
plt.figure(figsize=(10, 6))
plt.bar(data['Sample'], data['Fungal_Diversity'])
plt.title('Fungal Diversity on Lemon Surfaces')
plt.xlabel('Sample')
plt.ylabel('Diversity Index')
plt.show()
```
