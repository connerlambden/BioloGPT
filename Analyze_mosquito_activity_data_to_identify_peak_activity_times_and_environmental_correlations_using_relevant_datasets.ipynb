This notebook analyzes mosquito activity data to identify patterns and correlations with environmental factors.

```python
import pandas as pd

# Load mosquito activity data
data = pd.read_csv('mosquito_activity_data.csv')

# Analyze peak activity times
peak_activity = data.groupby('species')['activity_time'].max()
```

The analysis reveals peak activity times for different mosquito species.

```python
# Visualize the results
import matplotlib.pyplot as plt
peak_activity.plot(kind='bar')
plt.title('Peak Activity Times of Mosquito Species')
plt.xlabel('Species')
plt.ylabel('Peak Activity Time')
plt.show()
```
