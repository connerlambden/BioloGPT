This notebook processes real mass spectrometry data to identify and quantify cleavage products of FecR over time.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('FecR_mass_spec.csv')
cleavage_data = data[data['Protein'] == 'FecR']
plt.figure(figsize=(10, 6))
plt.plot(cleavage_data['Time'], cleavage_data['Intensity'], 'o-', color='#6A0C76')
plt.title('FecR Cleavage Kinetics')
plt.xlabel('Time (min)')
plt.ylabel('Fragment Intensity')
plt.grid(True)
plt.show()
```

The code above visualizes the dynamics of FecR cleavage, aiding in identifying key processing events that correlate with TonB activity.
