This notebook will analyze the performance metrics of microbioreactors, focusing on oxygen levels and cell density.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Flow Rate': ['0.0005 µl/sec', '0.22 µl/sec'], 'Cell Density': [8.01, 1.92], 'Oxygen Content': [1.76, 8.2]}
df = pd.DataFrame(data)

plt.bar(df['Flow Rate'], df['Oxygen Content'], color='blue', label='Oxygen Content (mg/L)')
plt.bar(df['Flow Rate'], df['Cell Density'], color='orange', label='Cell Density (cells/m^2)', alpha=0.5)
plt.title('Microbioreactor Performance Metrics')
plt.xlabel('Flow Rate')
plt.ylabel('Measurements')
plt.legend()
plt.show()
```

This analysis provides insights into how flow rates affect oxygen levels and cell density in microbioreactors.
