This notebook will analyze 5hmC levels in colorectal cancer samples, comparing primary tumors and liver metastases.

```python
import pandas as pd
# Load datasets
primary_tumors = pd.read_csv('primary_tumors.csv')
metastases = pd.read_csv('metastases.csv')
# Analyze 5hmC levels
primary_avg = primary_tumors['5hmC'].mean()
metastases_avg = metastases['5hmC'].mean()
print(f'Average 5hmC in primary tumors: {primary_avg}, in metastases: {metastases_avg}')
```
