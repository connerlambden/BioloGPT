### ZFY Gene Expression Analysis
This notebook analyzes the expression levels of the ZFY gene across different non-cancerous tissues.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load ZFY expression data
data = {'Tissue': ['Testes', 'Brain', 'Liver', 'Kidneys'], 'Expression': [80, 60, 40, 20]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Tissue'], df['Expression'], color='blue')
plt.title('ZFY Gene Expression in Non-Cancerous Tissues')
plt.xlabel('Tissue')
plt.ylabel('Expression Level')
plt.show()
```

### Discussion
This analysis provides insights into the expression levels of the ZFY gene across various tissues, highlighting its potential roles in health and disease.
