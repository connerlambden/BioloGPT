### Analysis of Marine-Derived Compounds
This notebook analyzes the bioactivity data of marine-derived compounds, focusing on their potential therapeutic applications.

```python
import pandas as pd

# Load bioactivity data
bioactivity_data = pd.read_csv('marine_compounds_bioactivity.csv')

# Analyze binding affinities
binding_affinities = bioactivity_data[['compound', 'binding_affinity']].sort_values(by='binding_affinity')

# Display results
binding_affinities
```
