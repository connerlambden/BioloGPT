### Análisis de Datos de Poliadenilación
Este análisis se centrará en identificar y comparar las señales de poliadenilación en diferentes tipos celulares utilizando datos de RNA-seq.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar datos de poliadenilación
data = pd.read_csv('data/polya_signals.csv')

# Visualizar la distribución de señales de poliadenilación
plt.figure(figsize=(10, 6))
sns.histplot(data['signal_strength'], bins=30, kde=True)
plt.title('Distribución de Señales de Poliadenilación')
plt.xlabel('Fuerza de la Señal')
plt.ylabel('Frecuencia')
plt.show()
```

### Resultados
Este gráfico muestra la distribución de las señales de poliadenilación, lo que puede ayudar a identificar patrones relevantes en diferentes tipos celulares.

```python
# Guardar resultados en un archivo
data.to_csv('results/polya_analysis_results.csv', index=False)
```
