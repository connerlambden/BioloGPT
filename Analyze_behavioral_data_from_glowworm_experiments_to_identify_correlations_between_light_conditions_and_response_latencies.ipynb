This notebook analyzes the impact of sALAN on glow-worm behavior using collected data.

```python
import pandas as pd

data = pd.read_csv('glow_worm_data.csv')

# Analyze response latencies under different light conditions
latency_analysis = data.groupby('light_condition')['latency'].mean()
```
