This notebook will analyze RNA-seq data to identify opsin gene expression patterns in Odonata.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load RNA-seq data
data = pd.read_csv('odonata_opsin_expression.csv')

# Visualize expression levels
sns.boxplot(x='gene', y='expression_level', data=data)
plt.title('Opsin Gene Expression in Odonata')
plt.show()
```

The boxplot above illustrates the expression levels of various opsin genes in Odonata, highlighting those associated with polarization sensitivity.
