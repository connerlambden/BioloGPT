### Analysis of Maturation Inhibitor Efficacy
This notebook analyzes the IC50 values of various maturation inhibitors against HIV-1 subtypes.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for analysis
data = {'Compound': ['CV-10237', 'CV-10238', 'CV-10239', 'CV-10240', 'CV-10241'],
        'IC50_K3016': [2.59, 3.03, 3.28, 3.21, 13.84]}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Compound'], df['IC50_K3016'], color='blue')
plt.title('IC50 Values of Maturation Inhibitors')
plt.xlabel('Compounds')
plt.ylabel('IC50 (nM)')
plt.show()
```

### Discussion
This analysis highlights the significant differences in IC50 values based on linker length, supporting the hypothesis that shorter linkers enhance efficacy.
