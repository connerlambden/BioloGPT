This notebook analyzes biological datasets to compute Pearson's and Spearman's correlation coefficients.

```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr

data = pd.read_csv('biological_data.csv')
pearson_corr, _ = pearsonr(data['variable1'], data['variable2'])
spearman_corr, _ = spearmanr(data['variable1'], data['variable2'])
print(f'Pearson correlation: {pearson_corr}, Spearman correlation: {spearman_corr}')
```

The results provide insights into the relationships between the selected variables.

```python
# Display results in a user-friendly format
print(f'Results: Pearson = {pearson_corr}, Spearman = {spearman_corr}')
```
