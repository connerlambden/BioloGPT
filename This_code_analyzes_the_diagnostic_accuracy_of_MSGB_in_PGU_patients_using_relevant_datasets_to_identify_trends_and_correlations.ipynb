### Analysis of MSGB Diagnostic Accuracy in PGU Patients
This notebook will analyze the sensitivity and specificity of MSGB in diagnosing ocular sarcoidosis in PGU patients using available datasets.

```python
import pandas as pd

# Load relevant datasets
dataset = pd.read_csv('msgb_pgu_data.csv')

# Analyze sensitivity and specificity
sensitivity = dataset['sensitivity'].mean()
specificity = dataset['specificity'].mean()

print(f'Average Sensitivity: {sensitivity}%')
print(f'Average Specificity: {specificity}%')
```
