This notebook analyzes the impact of temperature variations on bell pepper yield, utilizing datasets from relevant studies.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Temperature Condition': ['Optimal', 'Low Night', 'High Day'], 'Yield Impact (%)': [100, 50, 30]}
df = pd.DataFrame(data)

plt.bar(df['Temperature Condition'], df['Yield Impact (%)'], color='green')
plt.title('Impact of Temperature on Bell Pepper Yield')
plt.xlabel('Temperature Condition')
plt.ylabel('Yield Impact (%)')
plt.show()
```

The bar graph visualizes the impact of different temperature conditions on bell pepper yield, highlighting the critical need for temperature management.
