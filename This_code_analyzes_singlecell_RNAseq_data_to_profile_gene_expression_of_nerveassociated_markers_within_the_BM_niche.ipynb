The following notebook downloads BM single-cell RNA-seq datasets and identifies expression profiles for sympathetic neuronal markers and niche cytokines.

```python
import scanpy as sc
adata = sc.read('path_to_BM_scRNAseq_data.h5ad')
markers = ['TH', 'NGF', 'BDNF', 'S100B']
sc.pl.violin(adata, keys=markers, groupby='cell_type')
```

This visualization helps correlate nerve damage with changes in cytokine and signaling profiles within the hematopoietic niche.

```python
sc.tl.rank_genes_groups(adata, 'cell_type', method='wilcoxon')
sc.pl.rank_genes_groups_heatmap(adata, groupby='cell_type', n_genes=10)
```

The heatmap provides further insights into the differential gene expression associated with neuropathy and malignant transformation.
