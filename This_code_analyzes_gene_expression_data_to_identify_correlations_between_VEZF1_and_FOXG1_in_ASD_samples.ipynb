# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load gene expression data
as_data = pd.read_csv('asd_gene_expression.csv')
control_data = pd.read_csv('control_gene_expression.csv')

# Calculate correlations
correlation = as_data['VEZF1'].corr(as_data['FOXG1'])
