Analyze gene expression datasets from FMT studies to identify changes associated with <em>Pseudoscardovia</em> enrichment.

```python
import pandas as pd

# Load gene expression data
gene_data = pd.read_csv('gene_expression_fmt.csv')

# Analyze changes in immune-related genes
immune_genes = gene_data[gene_data['gene_type'] == 'immune']

# Summary statistics
summary = immune_genes.describe()
```
