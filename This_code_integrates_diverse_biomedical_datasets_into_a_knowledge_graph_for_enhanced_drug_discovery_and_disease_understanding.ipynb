### Step 1: Data Integration
Integrate various biomedical datasets into a unified knowledge graph.

```python
import pandas as pd
# Load datasets
protein_data = pd.read_csv('protein_data.csv')
disease_data = pd.read_csv('disease_data.csv')
# Merge datasets into a knowledge graph
knowledge_graph = pd.merge(protein_data, disease_data, on='common_column')
```

### Step 2: Graph Construction
Construct the knowledge graph using the integrated data.

```python
import networkx as nx
# Create a directed graph
G = nx.DiGraph()
# Add nodes and edges
for index, row in knowledge_graph.iterrows():
    G.add_node(row['protein'])
    G.add_node(row['disease'])
    G.add_edge(row['protein'], row['disease'])
```

### Step 3: Visualization
Visualize the constructed knowledge graph.

```python
import matplotlib.pyplot as plt
# Draw the graph
nx.draw(G, with_labels=True)
plt.show()
```
