This notebook analyzes the microhardness data of radicular dentine treated with Nanochitosan and calcium hydroxide.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for microhardness changes
data = {'Treatment': ['Nanochitosan', 'Calcium Hydroxide'], '1 Week': [-2.2, -9.1], '4 Weeks': [-10.2, -17.2]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Treatment'], df['1 Week'], color='blue', label='1 Week')
plt.bar(df['Treatment'], df['4 Weeks'], color='red', label='4 Weeks', alpha=0.5)
plt.title('Microhardness Changes Over Time')
plt.ylabel('Microhardness Change (%)')
plt.legend()
plt.show()
```

This analysis provides insights into the comparative effects of Nanochitosan and calcium hydroxide on dentine microhardness.
