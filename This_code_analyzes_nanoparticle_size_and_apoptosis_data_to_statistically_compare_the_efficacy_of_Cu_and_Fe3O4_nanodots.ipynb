This notebook loads the nanoparticle size and apoptotic response data from the study and creates visualizations to compare the effects of Cu and Fe3O4 nanodots.

```python
import pandas as pd
import plotly.express as px

# Nanoparticle size data
np_data = {'Nanodot_Type': ['Cu', 'Fe3O4'], 'Size (nm)': [4.7, 7]}
df_np = pd.DataFrame(np_data)
fig_np = px.bar(df_np, x='Nanodot_Type', y='Size (nm)', title='Nanodot Size Comparison')
fig_np.show()

# Apoptosis data for Cu nanodots in A549 cells
apoptosis_data = {'Concentration (ng/µL)': ['0.1', '1', '50', '100'], 'A549_pre_apoptosis (%)': [40, 91.4, 88, 87.7]}
df_apoptosis = pd.DataFrame(apoptosis_data)
fig_apoptosis = px.line(df_apoptosis, x='Concentration (ng/µL)', y='A549_pre_apoptosis (%)', markers=True, title='A549 Pre-Apoptosis by Cu Nanodots')
fig_apoptosis.show()
```
