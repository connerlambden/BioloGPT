Extract detailed binding interface metrics from crystallographic datasets of the M1R–nanobody complex.

```python
import pandas as pd
# Load and analyze binding affinity data
data = pd.read_csv('binding_data.csv')
print(data.describe())
```

Visualize interaction networks to identify key residues for binding optimization.

```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
G.add_edge('M1R', 'M1R-01', weight=0.9)
plt.figure(figsize=(6,6))
nx.draw_networkx(G, with_labels=True)
plt.show()
```
