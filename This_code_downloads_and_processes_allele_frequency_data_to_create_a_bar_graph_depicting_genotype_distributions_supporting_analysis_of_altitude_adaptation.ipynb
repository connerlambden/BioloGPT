Load allele frequency dataset and prepare it for visualization.

```python
import pandas as pd
import plotly.express as px

data = pd.DataFrame([
    {'genotype': 'II', 'I_allele': 0.57, 'D_allele': 0.43},
    {'genotype': 'ID', 'I_allele': 0.32, 'D_allele': 0.68},
    {'genotype': 'DD', 'I_allele': 0.30, 'D_allele': 0.70}
])
fig = px.bar(data, x='genotype', y=['I_allele', 'D_allele'], barmode='group', title='Allele Frequencies by Genotype')
fig.show()
```
