## Differential Gene Expression Analysis of Mosquito Species Under Interspecific Competition

This notebook analyzes RNA-seq data to identify genes associated with competitive interactions between Aedes aegypti and Anopheles species.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
df = pd.read_csv('Aedes_Anopheles_competition_data.csv')
```

### Step 1: Load and Inspect Data

We will start by loading the dataset and inspecting its structure to understand the variables involved in the analysis.

```python
print(df.head())
print(df.describe())
print(df.info())
```

### Step 2: Normalize Gene Expression Data

Normalization is crucial for accurate comparison of gene expression levels across samples.

```python
# Assuming gene expression counts are in 'counts' column
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['normalized_counts'] = scaler.fit_transform(df[['counts']])
```

### Step 3: Differential Expression Analysis

Identify genes that are differentially expressed between species under competitive conditions.

```python
group_a = df[df['species'] == 'Aedes aegypti']['normalized_counts']
group_b = df[df['species'] == 'Anopheles']['normalized_counts']
t_stat, p_val = stats.ttest_ind(group_a, group_b)
print(f'T-test statistic: {t_stat}, p-value: {p_val}')
```

### Step 4: Visualization of Differential Expression

Visualize the differences in gene expression between the two species.

```python
sns.boxplot(x='species', y='normalized_counts', data=df)
plt.title('Gene Expression Comparison')
plt.show()
```

### Step 5: Interpretation of Results

Interpret the statistical results to understand the biological significance of differentially expressed genes.

```python
if p_val < 0.05:
    print('Significant differential expression observed.')
else:
    print('No significant differential expression.')
```

## Conclusion

The analysis highlights key genes that may be responsible for competitive advantages in Aedes aegypti over Anopheles species.
