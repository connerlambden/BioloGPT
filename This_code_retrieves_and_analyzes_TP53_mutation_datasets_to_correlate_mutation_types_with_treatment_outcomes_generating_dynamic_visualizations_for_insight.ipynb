This notebook details downloading TP53 mutation data for NSCLC and performing statistical analysis on chemo- and radiation-response.

```python
import pandas as pd
import plotly.express as px
url = 'URL_to_TP53_dataset'
df = pd.read_csv(url)
fig = px.bar(df, x='Mutation_Type', y='Percentage', title='TP53 Mutation Spectrum in NSCLC')
fig.show()
```

Next, the notebook applies t-tests to evaluate the differences in drug sensitivity between mutant and wild-type TP53 cases.

```python
from scipy import stats
stat, pvalue = stats.ttest_ind(df[df['Mutation_Status']=='Mutant']['Gemcitabine_Sensitivity'], df[df['Mutation_Status']=='Wild-Type']['Gemcitabine_Sensitivity'])
print(f'Statistic: {stat}, p-value: {pvalue}')
```
