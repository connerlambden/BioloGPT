### Sample Size Calculation for Vaccine Trials
This notebook calculates the required sample size for a vaccine trial based on user-defined parameters.

```python
import scipy.stats as stats

def calculate_sample_size(effect_size, alpha=0.05, power=0.8):
    z_alpha = stats.norm.ppf(1 - alpha / 2)
    z_beta = stats.norm.ppf(power)
    n = ((z_alpha + z_beta) ** 2) / (effect_size ** 2)
    return n

# Example usage
calculate_sample_size(effect_size=0.05)  # Adjust effect size as needed.
```

### Discussion
This function provides a simple way to estimate sample size based on the expected effect size, significance level, and desired power.

```python
# Example of calculating sample size
sample_size = calculate_sample_size(0.05)
print(f'Required sample size: {sample_size}')
```
