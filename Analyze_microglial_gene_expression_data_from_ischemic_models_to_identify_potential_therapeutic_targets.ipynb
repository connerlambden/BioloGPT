This notebook will analyze gene expression data from microglia in ischemic conditions, focusing on identifying markers associated with M1 and M2 polarization.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load microglial gene expression data
gene_data = pd.read_csv('microglia_expression.csv')

# Analyze expression levels of key markers
m1_markers = gene_data[gene_data['marker'].isin(['CD11b', 'CD68'])]
m2_markers = gene_data[gene_data['marker'].isin(['Ym1', 'CD206'])]

# Plotting the expression levels
plt.figure(figsize=(10, 6))
plt.bar(m1_markers['time_point'], m1_markers['expression_level'], label='M1 Markers')
plt.bar(m2_markers['time_point'], m2_markers['expression_level'], label='M2 Markers', alpha=0.7)
plt.xlabel('Time Point')
plt.ylabel('Expression Level')
plt.title('Microglial Marker Expression Over Time')
plt.legend()
plt.show()
```

This analysis provides insights into the temporal dynamics of microglial activation and can help identify critical time points for therapeutic intervention.
