### Nucleotide Diversity and Selection Analysis of *C. quadrata*

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for nucleotide diversity and selection
data = {
    'gene': ['ccsA', 'psbF', 'chlN', 'cemA', 'rps3', 'rps18'],
    'nucleotide_diversity': [0.15, 0.15, 0.15, 0.15, 0.15, 0.15],
    'selection': ['positive', 'positive', 'positive', 'positive', 'positive', 'positive']
}
df = pd.DataFrame(data)

# Plotting nucleotide diversity
plt.figure(figsize=(10, 6))
plt.bar(df['gene'], df['nucleotide_diversity'], color='skyblue')
plt.title('Nucleotide Diversity in Chloroplast Genes of *C. quadrata*')
plt.xlabel('Gene')
plt.ylabel('Nucleotide Diversity')
plt.axhline(y=0.15, color='r', linestyle='--', label='Selection Threshold')
plt.legend()
plt.show()
```
