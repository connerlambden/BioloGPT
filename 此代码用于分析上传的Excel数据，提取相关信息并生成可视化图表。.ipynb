```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取Excel文件
data = pd.read_excel('data.xlsx')

# 数据处理
summary = data.describe()

# 生成可视化图表
plt.figure(figsize=(10, 6))
plt.plot(data['Column1'], data['Column2'])
plt.title('数据可视化')
plt.xlabel('Column1')
plt.ylabel('Column2')
plt.show()
```
