### Data Analysis of NDV Propagation in Vero vs. BHK-21 Cells

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
vero_data = pd.read_csv('vero_ndv_data.csv')
bhk21_data = pd.read_csv('bhk21_ndv_data.csv')

# Analyze and visualize data
plt.figure(figsize=(10, 5))
plt.plot(vero_data['passage'], vero_data['viral_yield'], label='Vero Cells', marker='o')
plt.plot(bhk21_data['passage'], bhk21_data['viral_yield'], label='BHK-21 Cells', marker='x')
plt.title('NDV Propagation in Vero vs. BHK-21 Cells')
plt.xlabel('Passage Number')
plt.ylabel('Viral Yield')
plt.legend()
plt.show()
```
