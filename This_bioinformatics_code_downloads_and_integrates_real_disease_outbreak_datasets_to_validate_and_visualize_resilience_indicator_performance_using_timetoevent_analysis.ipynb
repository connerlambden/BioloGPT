The notebook demonstrates dataset integration and the implementation of survival analysis to compute lead times from outbreak data.

```python
import pandas as pd
import lifelines
import plotly.express as px

# Downloading the outbreak dataset from GitHub
url = 'https://github.com/QZhao16/EWSs.across.diseases/blob/main/data.csv'
df = pd.read_csv(url)

# Applying survival analysis using lifelines
from lifelines import CoxPHFitter
cph = CoxPHFitter()
cph.fit(df, duration_col='lead_time', event_col='outbreak_event')
cph.print_summary()

# Visualizing the survival function
fig = px.line(df, x='date', y='lead_time', title='Outbreak Lead Time Analysis')
fig.show()
```
