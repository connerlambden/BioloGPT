This notebook will analyze the impact of fiber size on thermal conductivity in PVA-based wood composites using relevant datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
fiber_sizes = [0.4, 1.0, 2.0]
thermal_conductivities = [0.2, 0.18, 0.14]

# Create DataFrame
data = pd.DataFrame({'Fiber Size (mm)': fiber_sizes, 'Thermal Conductivity (W/m·K)': thermal_conductivities})

# Plotting
plt.bar(data['Fiber Size (mm)'], data['Thermal Conductivity (W/m·K)'])
plt.title('Thermal Conductivity vs Fiber Size')
plt.xlabel('Fiber Size (mm)')
plt.ylabel('Thermal Conductivity (W/m·K)')
plt.show()
```

This analysis provides insights into how fiber size affects thermal conductivity, which is crucial for optimizing insulation materials.
