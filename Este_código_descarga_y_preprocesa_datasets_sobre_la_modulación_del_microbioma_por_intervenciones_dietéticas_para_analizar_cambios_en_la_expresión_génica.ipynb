Cargar datasets relevantes, por ejemplo, E-GEOD-25161, para analizar perfiles de expresión asociados a intervenciones nutracéuticas en el microbioma.

```python
import pandas as pd
url = 'https://www.ebi.ac.uk/biostudies/files/E-GEOD-25161/E-GEOD-25161.json'
data = pd.read_json(url)
print(data.head())
```
