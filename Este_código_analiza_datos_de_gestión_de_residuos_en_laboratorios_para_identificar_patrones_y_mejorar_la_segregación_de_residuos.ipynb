Análisis de datos de gestión de residuos en laboratorios.

```python
import pandas as pd

# Cargar datos de residuos
data = pd.read_csv('residuos_laboratorio.csv')

# Análisis de segregación
data.groupby('tipo_residuo').size()
```

Este análisis permite identificar qué tipos de residuos son más comunes y dónde se pueden mejorar las prácticas de segregación.

```python
# Visualización de los resultados
import matplotlib.pyplot as plt

residuo_counts = data['tipo_residuo'].value_counts()
residuo_counts.plot(kind='bar')
plt.title('Distribución de Tipos de Residuos')
plt.xlabel('Tipo de Residuo')
plt.ylabel('Cantidad')
plt.show()
```
