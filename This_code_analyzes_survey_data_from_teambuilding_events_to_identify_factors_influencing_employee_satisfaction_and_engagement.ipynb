### Team Building Survey Data Analysis
This notebook analyzes survey data collected from team-building events to identify key factors that influence employee satisfaction and engagement.

```python
import pandas as pd

# Load survey data
data = pd.read_csv('team_building_survey.csv')

# Display the first few rows of the dataset
print(data.head())
```

### Data Analysis
We will analyze the data to find correlations between different activities and employee satisfaction.

```python
# Analyze satisfaction levels based on different activities
satisfaction_analysis = data.groupby('activity')['satisfaction'].mean()
print(satisfaction_analysis)
```
