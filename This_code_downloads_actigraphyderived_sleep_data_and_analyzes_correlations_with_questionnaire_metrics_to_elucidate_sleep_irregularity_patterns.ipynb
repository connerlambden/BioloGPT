Retrieve and parse actigraphy data linked with sleep assessment scores from the provided dataset.

```python
import pandas as pd
import numpy as np

data = pd.read_csv('actigraphy_sleep_data.csv')
correlations = data[['sleep_irregularity', 'PSQI', 'DBAS', 'TUQ']].corr()
print(correlations)
```

This markdown explains the step-by-step process of loading, cleaning, and examining correlations between objective and subjective sleep measures.

```python
import plotly.graph_objects as go
fig = go.Figure(data=[go.Bar(x=correlations.columns, y=correlations.loc['sleep_irregularity'])])
fig.update_layout(title='Correlation of Sleep Irregularity with Sleep Measures', xaxis_title='Metrics', yaxis_title='Correlation Coefficient')
fig.show()
```
