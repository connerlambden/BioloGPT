Step-by-step analysis of electrophysiology and transcriptomics data from Tsc1 mutant versus control Purkinje neurons.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Download data from provided EBI dataset URL
url = 'https://www.ebi.ac.uk/arrayexpress/experiments/E-GEOD-61591'
data = pd.read_csv(url)
# Perform analysis correlating Nav current reduction with alteration in AIS protein expression
plt.figure(figsize=(8,5))
plt.scatter(data['Nav_current'], data['AnkyrinG_intensity'], color='green')
plt.xlabel('Nav Current Amplitude')
plt.ylabel('AnkyrinG Intensity')
plt.title('Correlation of Nav Current with AIS Integrity')
plt.show()
```

This notebook provides a clear analysis pipeline to understand the electrophysiological impact of Tsc1 deletion.

```python
# Continue analysis
# Compute correlation coefficient
df = pd.DataFrame(data, columns=['Nav_current','AnkyrinG_intensity'])
correlation = df.corr().loc['Nav_current','AnkyrinG_intensity']
print('Correlation between Nav current and AnkyrinG expression:', correlation)
```
