Step-by-step analysis of RNA-seq and imaging data from TS zebrafish across temperature treatments.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load expression and phenotype datasets
expression_data = pd.read_csv('rna_seq_data.csv')
phenotype_data = pd.read_csv('phenotype_data.csv')
# Merge and analyze data
merged_data = pd.merge(expression_data, phenotype_data, on='sample_id')
plt.scatter(merged_data['temperature'], merged_data['arrhythmia_score'])
plt.xlabel('Temperature (°C)')
plt.ylabel('Arrhythmia Score')
plt.title('Relationship Between Temperature and Cardiac Dysfunction')
plt.show()
```

Analysis reveals the correlation between temperature elevation and arrhythmic events, supporting the hypothesis.

```python
# Further statistical analysis
import scipy.stats as stats
corr_coef, p_value = stats.pearsonr(merged_data['temperature'], merged_data['arrhythmia_score'])
print('Correlation coefficient:', corr_coef)
print('P-value:', p_value)
```
