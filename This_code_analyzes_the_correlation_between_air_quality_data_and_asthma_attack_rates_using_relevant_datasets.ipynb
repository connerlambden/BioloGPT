### Data Analysis of Air Quality and Asthma Attacks

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('asthma_air_quality_data.csv')

plt.scatter(data['Air Quality Index'], data['Asthma Attacks'])
plt.title('Correlation between Air Quality and Asthma Attacks')
plt.xlabel('Air Quality Index')
plt.ylabel('Number of Asthma Attacks')
plt.show()
```
