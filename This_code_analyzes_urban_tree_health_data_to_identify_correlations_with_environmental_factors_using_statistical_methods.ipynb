### Urban Tree Health Analysis
This notebook will analyze the health status of spherical crown taxa in Budapest in relation to environmental factors.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('urban_tree_health_data.csv')

# Analyze correlations
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix of Urban Tree Health Factors')
plt.show()
```
