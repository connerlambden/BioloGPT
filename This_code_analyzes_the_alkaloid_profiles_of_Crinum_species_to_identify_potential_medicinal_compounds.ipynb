### Alkaloid Profile Analysis of Crinum Species

```python
import pandas as pd

# Load alkaloid data
alkaloid_data = pd.read_csv('crinum_alkaloids.csv')

# Display the data
alkaloid_data.head()
```

### Statistical Analysis of Alkaloid Activity

```python
# Perform statistical analysis on the alkaloid activity
activity_summary = alkaloid_data.describe()
activity_summary
```

### Visualization of Alkaloid Activity

```python
import matplotlib.pyplot as plt

# Plotting the alkaloid activity
plt.bar(alkaloid_data['species'], alkaloid_data['activity'])
plt.xlabel('Species')
plt.ylabel('Activity')
plt.title('Alkaloid Activity in Crinum Species')
plt.show()
```
