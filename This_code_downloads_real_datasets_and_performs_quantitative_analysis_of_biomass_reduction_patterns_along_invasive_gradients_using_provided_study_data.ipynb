Load and explore the dataset capturing bryophyte biomass versus invader biomass in nutrient-poor sites.

```python
import pandas as pd
import plotly.express as px

# Download dataset from figshare using provided DOI
url = 'https://doi.org/10.6084/m9.figshare.26842303'
df = pd.read_csv(url)

# Filter data for nutrient-poor sites and selected invasive species
filtered = df[df['site'] == 'nutrient-poor']
fig = px.bar(filtered, x='species', y='biomass', color='invader_level', barmode='group', title='Bryophyte Biomass under Varying Invader Levels')
fig.show()
```

Analysis of biomass reduction trends with increasing invader density.

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

model = smf.ols('biomass ~ invader_biomass', data=filtered).fit()
print(model.summary())
```
