Load the clinical dataset containing QFM levels and relevant patient clinical parameters. This data supports visualization and further analysis.

```python
import pandas as pd
import plotly.express as px

# Load dataset
# data = pd.read_csv('clinical_data.csv')

data = pd.DataFrame({
    'Group': ['Non-HCAI', 'HCAI'],
    'D1_QFM': [5.75, 4.40],
    'D2_QFM': [6.10, 4.38]
})

# Create bar chart
fig = px.bar(data, x='Group', y=['D1_QFM', 'D2_QFM'], barmode='group', title='QFM Levels by Group')
fig.show()
```

This notebook demonstrates how to visualize differences in QFM levels between patient groups, aiding in validation of its predictive utility for HCAI.
