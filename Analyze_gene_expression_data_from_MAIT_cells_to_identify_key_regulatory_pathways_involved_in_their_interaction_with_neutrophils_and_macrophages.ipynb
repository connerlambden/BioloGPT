This notebook will analyze RNA-seq data from MAIT cells to identify differentially expressed genes during bacterial infections.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('mait_cells_expression.csv')

# Analyze differential expression
# (Assuming the dataset has columns for gene names and expression levels)
differential_expression = data[data['condition'] == 'infected'] - data[data['condition'] == 'control']

# Visualize results
plt.figure(figsize=(10, 6))
sns.barplot(x='gene', y='expression', data=differential_expression)
plt.title('Differential Expression of MAIT Cells in Infection')
plt.xticks(rotation=90)
plt.show()
```

This analysis will help identify which genes are upregulated in MAIT cells during bacterial infections, providing insights into their regulatory mechanisms.
