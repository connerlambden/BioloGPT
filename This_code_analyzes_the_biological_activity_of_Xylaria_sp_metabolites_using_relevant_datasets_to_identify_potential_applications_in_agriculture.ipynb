This notebook explores the biological activities of Xylaria sp. metabolites, focusing on their phytotoxic and stimulatory effects.

```python
import pandas as pd

# Load relevant datasets
datasets = pd.read_csv('xylaria_metabolites.csv')

# Analyze the effects of metabolites on Lactuca sativa
results = datasets.groupby('compound').agg({'effect': 'mean'})
results
```

The analysis reveals the average effects of each compound on plant growth, highlighting those with dual actions.

```python
results.to_csv('xylaria_effects_analysis.csv') # Save the results for further analysis.
```
