### Data Analysis of Diastolic Pressure and Blood Flow
This notebook will analyze patient data to explore the relationship between diastolic pressure and blood flow.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load patient data
data = pd.read_csv('patient_data.csv')

# Analyze the relationship between diastolic pressure and blood flow
plt.scatter(data['diastolic_pressure'], data['blood_flow'])
plt.title('Diastolic Pressure vs Blood Flow')
plt.xlabel('Diastolic Pressure (mmHg)')
plt.ylabel('Blood Flow (mL/min)')
plt.show()
```
