### Data Analysis of Stickies Size Distribution
This analysis will utilize data from the heat-dispersing study to evaluate how temperature and pulp consistency affect the size of stickies.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data based on the extracted data
data = {
    'Temperature (°C)': [70, 80, 90, 100],
    'Size (mm²)': [0.04, 0.038, 0.034, 0.028],
    'Removal (%)': [20.1, 21.3, 24.8, 25.7]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Temperature (°C)'], df['Size (mm²)'], marker='o')
plt.title('Effect of Temperature on Size of Stickies')
plt.xlabel('Temperature (°C)')
plt.ylabel('Size (mm²)')
plt.grid()
plt.show()
```
