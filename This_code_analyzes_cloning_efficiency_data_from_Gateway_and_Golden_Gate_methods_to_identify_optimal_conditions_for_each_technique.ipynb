### Cloning Efficiency Analysis
This notebook will analyze the efficiency of Gateway and Golden Gate cloning methods based on experimental data.

```python
import pandas as pd

data = {'Method': ['Gateway', 'Golden Gate'], 'Efficiency': [75, 95]}
df = pd.DataFrame(data)

# Calculate average efficiency
avg_efficiency = df['Efficiency'].mean()
print(f'Average Cloning Efficiency: {avg_efficiency}%')
```

### Results
The analysis shows the average cloning efficiency across both methods, providing insights into their comparative performance.
