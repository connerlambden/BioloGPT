Análisis de expresión génica en el citoplasma utilizando datos de RNA-seq.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar datos de expresión génica
expression_data = pd.read_csv('data.csv')

# Visualizar datos
sns.boxplot(data=expression_data)
plt.title('Expresión génica en el citoplasma')
plt.show()
```
