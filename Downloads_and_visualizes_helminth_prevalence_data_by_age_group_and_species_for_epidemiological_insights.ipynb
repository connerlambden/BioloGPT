This notebook loads the real data extracted from the study and generates interactive bar and pie charts to visualize variations in parasite prevalence among different age groups and among species.

```python
import pandas as pd
import plotly.express as px

# Data for age groups
age_data = pd.DataFrame([
    {'Age Group': '8 years', 'Prevalence': 41.94},
    {'Age Group': '9 years', 'Prevalence': 34.61},
    {'Age Group': '10 years', 'Prevalence': 23.65},
    {'Age Group': '11 years', 'Prevalence': 19.56}
])

fig_age = px.bar(age_data, x='Age Group', y='Prevalence', title='Helminth Prevalence by Age Group', labels={'Prevalence': 'Prevalence (%)'})
fig_age.show()

# Data for helminth species
species_data = pd.DataFrame([
    {'Species': 'Dujardinascaris', 'Prevalence': 18.01},
    {'Species': 'Trichinella', 'Prevalence': 5.28},
    {'Species': 'Polydeliphs', 'Prevalence': 3.1},
    {'Species': 'Renifers', 'Prevalence': 0.62}
])

fig_species = px.pie(species_data, names='Species', values='Prevalence', title='Helminth Species Prevalence')
fig_species.show()
```
