### Neuronal Data Analysis
This notebook analyzes neuronal data from various brain regions to identify patterns of entrainment.

```python
import pandas as pd

# Sample data
data = {
    'Region': ['Cortex', 'Thalamus', 'Striatum', 'BLA'],
    'Number of Neurons': [1307, 1705, 303, 573],
    'Number of Rats': [9, 10, 7, 7]
}

df = pd.DataFrame(data)

# Display the dataframe
df
```

### Data Visualization
Visualizing the neuronal data to identify trends across brain regions.

```python
import matplotlib.pyplot as plt

# Bar plot
plt.bar(df['Region'], df['Number of Neurons'], color='blue', alpha=0.6, label='Number of Neurons')
plt.bar(df['Region'], df['Number of Rats'], color='orange', alpha=0.6, label='Number of Rats')
plt.title('Neuronal Data Across Brain Regions')
plt.xlabel('Brain Region')
plt.ylabel('Count')
plt.legend()
plt.show()
```
