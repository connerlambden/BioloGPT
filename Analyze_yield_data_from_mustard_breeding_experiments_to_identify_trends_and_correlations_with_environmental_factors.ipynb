This notebook analyzes mustard yield data to understand the impact of breeding on crop performance.

```python
import pandas as pd
import matplotlib.pyplot as plt

yield_data = pd.DataFrame({
    'Year': [1979, 1980, 1981, 1982, 1990, 1997, 1998, 1999, 2000],
    'B_juncea_yield': [122, 86, 35, 40, 107, 177, 242, 209, 117],
    'B_napus_yield': [187, 33, 56, 11, 0, 0, 0, 0, 0]
})

plt.figure(figsize=(10, 5))
plt.bar(yield_data['Year'], yield_data['B_juncea_yield'], label='B. juncea')
plt.bar(yield_data['Year'], yield_data['B_napus_yield'], label='B. napus')
plt.xlabel('Year')
plt.ylabel('Yield (kg/ha)')
plt.title('Mustard Yield Comparison')
plt.legend()
plt.show()
```

This analysis provides insights into the yield performance of different mustard varieties over the years.
