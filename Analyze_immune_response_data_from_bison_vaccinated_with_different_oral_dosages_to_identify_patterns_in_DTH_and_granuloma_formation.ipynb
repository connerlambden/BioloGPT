This notebook analyzes immune response data from bison vaccinated with varying oral dosages of BCG.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('bison_vaccine_data.csv')

# Analyze DTH responses
plt.figure(figsize=(10, 6))
plt.scatter(data['dosage'], data['DTH_response'], color='blue')
plt.title('DTH Response vs. Vaccine Dosage')
plt.xlabel('Vaccine Dosage (mg)')
plt.ylabel('DTH Response (mm)')
plt.show()
```

This analysis visualizes the relationship between vaccine dosage and DTH response in bison.

```python
# Further analysis on granuloma formation
plt.figure(figsize=(10, 6))
plt.bar(data['dosage'], data['granuloma_count'], color='green')
plt.title('Granuloma Count vs. Vaccine Dosage')
plt.xlabel('Vaccine Dosage (mg)')
plt.ylabel('Granuloma Count')
plt.show()
```
