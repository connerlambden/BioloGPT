### Genetic Data Analysis of Sea Urchins
This notebook analyzes genetic data to compare adaptations in brooding and non-brooding sea urchins.

```python
import pandas as pd

# Load genetic data
brooding_data = pd.read_csv('brooding_urchins_genetic_data.csv')
non_brooding_data = pd.read_csv('non_brooding_urchins_genetic_data.csv')

# Analyze genetic diversity
brooding_diversity = brooding_data['genetic_diversity'].mean()
non_brooding_diversity = non_brooding_data['genetic_diversity'].mean()

# Compare results
print(f'Brooding Genetic Diversity: {brooding_diversity}')
print(f'Non-Brooding Genetic Diversity: {non_brooding_diversity}')
```
