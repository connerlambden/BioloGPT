### Acute Toxicity Analysis of Neomysis japonica
This notebook analyzes the acute toxicity data of Neomysis japonica to various pollutants, including heavy metals and microplastics.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for heavy metal toxicity
heavy_metal_data = {
    'Heavy Metal': ['Cadmium', 'Copper', 'Zinc', 'Arsenic', 'Lead', 'Chromium'],
    '96 hr LC50 (µg/L)': [90.4, 82.3, 646.1, 1488.5, 1758.8, 1609.6]
}

# Create DataFrame
df_heavy_metal = pd.DataFrame(heavy_metal_data)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df_heavy_metal['Heavy Metal'], df_heavy_metal['96 hr LC50 (µg/L)'], color='skyblue')
plt.title('Acute Toxicity of Heavy Metals on Neomysis japonica')
plt.xlabel('Heavy Metal')
plt.ylabel('96 hr LC50 (µg/L)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
