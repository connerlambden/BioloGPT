This section downloads the curated dataset and plots a comparative bar graph of both total bryophyte taxa and endemic taxa for each island.

```python
import plotly.graph_objects as go

# Define data for islands
data = {
    'islands': ['Azores', 'Madeira', 'São Tomé and Príncipe', 'Cape Verde', 'Canary Islands', 'Fernando de Noronha', 'Trindade', 'Ascension', 'Saint Helena', 'Tristan da Cunha'],
    'bryophyte_taxa': [471, 586, 292, 207, 522, 29, 35, 84, 101, 303],
    'endemic_taxa': [37, 47, 19, 8, 47, 1, 0, 15, 27, 51]
}

# Create bar chart
fig = go.Figure(data=[
    go.Bar(name='Bryophyte Taxa', x=data['islands'], y=data['bryophyte_taxa'], marker_color='#6A0C76'),
    go.Bar(name='Endemic Taxa', x=data['islands'], y=data['endemic_taxa'], marker_color='#b565a7')
])
fig.update_layout(title='Bryophyte Diversity and Endemism on Atlantic Islands', barmode='group', xaxis_title='Island', yaxis_title='Count')
fig.show()
```

The above code uses Plotly to generate an interactive grouped bar chart, summarizing key diversity metrics across islands.
