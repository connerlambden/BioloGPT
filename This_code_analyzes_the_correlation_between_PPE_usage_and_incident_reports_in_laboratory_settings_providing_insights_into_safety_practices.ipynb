### Analyzing PPE Usage and Incident Reports
This analysis aims to correlate PPE usage with the frequency of incident reports in laboratory settings.

```python
import pandas as pd

# Load data
data = pd.read_csv('lab_safety_data.csv')

# Analyze correlation
correlation = data['PPE_Usage'].corr(data['Incident_Reports'])
print(f'Correlation between PPE usage and incident reports: {correlation}')
```

### Discussion
The results will help determine if increased PPE usage correlates with fewer incidents, guiding future safety protocols.

```python
# Visualization of results
import matplotlib.pyplot as plt
plt.scatter(data['PPE_Usage'], data['Incident_Reports'])
plt.title('PPE Usage vs Incident Reports')
plt.xlabel('PPE Usage (%)')
plt.ylabel('Incident Reports')
plt.show()
```
