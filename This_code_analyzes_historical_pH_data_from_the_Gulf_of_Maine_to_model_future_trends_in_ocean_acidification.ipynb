### Historical pH Data Analysis for the Gulf of Maine

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load historical pH data
pH_data = pd.read_csv('gulf_of_maine_ph_data.csv')

# Plotting the pH over time
plt.figure(figsize=(10,5))
plt.plot(pH_data['Year'], pH_data['pH'], marker='o')
plt.title('Historical pH Levels in the Gulf of Maine')
plt.xlabel('Year')
plt.ylabel('pH Level')
plt.grid()
plt.show()
```

### This analysis provides insights into the trends of ocean acidification in the Gulf of Maine.

```python
# Future projections based on current trends
future_years = range(2025, 2051)
projected_ph = [pH_data['pH'].iloc[-1] - 0.02 * (year - 2025) for year in future_years]

# Plotting future projections
plt.figure(figsize=(10,5))
plt.plot(future_years, projected_ph, marker='x', color='red')
plt.title('Projected pH Levels in the Gulf of Maine (2025-2050)')
plt.xlabel('Year')
plt.ylabel('Projected pH Level')
plt.grid()
plt.show()
```
