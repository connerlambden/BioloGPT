This notebook will analyze the prevalence of AMR pathogens in NICUs using data from recent studies.

```python
import pandas as pd

# Load AMR pathogen data
amr_data = pd.read_csv('amr_pathogen_data.csv')

# Analyze prevalence
prevalence = amr_data.groupby('pathogen').agg({'cases':'sum'})
print(prevalence)
```
