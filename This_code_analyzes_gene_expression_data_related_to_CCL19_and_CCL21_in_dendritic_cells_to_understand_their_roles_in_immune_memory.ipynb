This notebook will analyze gene expression data from dendritic cells treated with CCL19 and CCL21.

```python
import pandas as pd

# Load gene expression data
data = pd.read_csv('gene_expression_data.csv')

# Analyze the expression levels of CCL19 and CCL21
ccl19_expression = data[data['gene'] == 'CCL19']
ccl21_expression = data[data['gene'] == 'CCL21']

# Visualize the results
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.bar(ccl19_expression['condition'], ccl19_expression['expression'], label='CCL19')
plt.bar(ccl21_expression['condition'], ccl21_expression['expression'], label='CCL21', alpha=0.7)
plt.xlabel('Condition')
plt.ylabel('Expression Level')
plt.title('CCL19 and CCL21 Expression Levels')
plt.legend()
plt.show()
```

This analysis provides insights into how CCL19 and CCL21 influence dendritic cell function.
