The notebook begins with downloading and preprocessing expression data from resistant and susceptible grapevine cultivars.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Code to load transcriptomic datasets and analyze expression levels of immune-related genes
expression_data = pd.read_csv('grapevine_expression.csv')
selected_genes = expression_data[['Gene', 'V_riparia', 'V_vinifera']]
selected_genes.plot(kind='bar', x='Gene', figsize=(10,6))
plt.title('Comparative Gene Expression in Grapevine Cultivars')
plt.xlabel('Gene')
plt.ylabel('Expression Level')
plt.show()
```

Subsequent sections delve into network analysis of the identified immune genes using correlation metrics.

```python
import networkx as nx
import matplotlib.pyplot as plt
# Create a graph of gene interactions based on correlation
G = nx.Graph()
# Example data: add nodes and edges (replace with actual analysis)
G.add_node('GeneA')
G.add_node('GeneB')
G.add_edge('GeneA', 'GeneB', weight=0.85)
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_color='#6A0C76', node_size=500)
nx.draw_networkx_edges(G, pos, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_color='white')
plt.title('Immune Gene Interaction Network')
plt.show()
```

The final section summarizes statistical tests comparing immune gene expression between the two species.

```python
import scipy.stats as stats
# Perform t-test between expression levels in Vitis riparia and Vitis vinifera
t_stat, p_val = stats.ttest_ind(expression_data['V_riparia'], expression_data['V_vinifera'])
print('T-statistic:', t_stat, 'P-value:', p_val)
```
