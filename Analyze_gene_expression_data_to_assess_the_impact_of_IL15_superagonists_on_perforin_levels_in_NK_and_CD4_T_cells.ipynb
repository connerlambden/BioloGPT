This notebook will analyze RNA-seq data from patients treated with IL-15 superagonists and anti-PD-1 to evaluate perforin expression.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('gene_expression_data.csv')

# Filter for perforin expression
data_perforin = data[data['gene'] == 'PRF1']

# Plot the expression levels
plt.figure(figsize=(10,6))
sns.boxplot(x='treatment', y='expression', data=data_perforin)
plt.title('Perforin Expression by Treatment Group')
plt.show()
```

This analysis will help determine if the combination therapy significantly increases perforin expression compared to controls.

```python
# Further analysis can include statistical tests to compare groups
from scipy import stats

# Perform t-test between groups
group1 = data_perforin[data_perforin['treatment'] == 'IL-15 + anti-PD-1']['expression']
group2 = data_perforin[data_perforin['treatment'] == 'anti-PD-1']['expression']
t_stat, p_value = stats.ttest_ind(group1, group2)
print(f'T-test results: t-statistic = {t_stat}, p-value = {p_value}')
```
