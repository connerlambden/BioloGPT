Step 1: Load gene expression data for the 490 identified gland genes from the iBeetle screen.

```python
import pandas as pd
url = 'https://example.org/dataset_stink_gland.csv'  # Replace with actual URL
gene_data = pd.read_csv(url)
gene_data.head()
```

Step 2: Perform gene ontology enrichment using gseapy to identify key pathways involved in gland function.

```python
import gseapy as gp

gene_list = gene_data['gene_id'].tolist()

enrichr_results = gp.enrichr(gene_list=gene_list, description='StinkGland', gene_sets='KEGG_2019_Human', cutoff=0.5)
enrichr_results.results.head()
```

Step 3: Visualize the top enriched KEGG pathways with Plotly.

```python
import plotly.express as px

enrich_df = enrichr_results.results.head(10)
fig = px.bar(enrich_df, x='Adjusted P-value', y='Term', orientation='h', title='Top KEGG Pathways')
fig.show()
```
