This notebook explores the themes of chance and human agency in Polybius' Histories.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load historical text data
text_data = pd.read_csv('polybius_histories.csv')

# Analyze themes
chance_count = text_data['text'].str.contains('tychê').sum()
human_agency_count = text_data['text'].str.contains('human agency').sum()

# Visualize results
plt.bar(['Chance', 'Human Agency'], [chance_count, human_agency_count])
plt.title('Themes in Polybius Histories')
plt.show()
```

This analysis reveals the frequency of themes related to chance and human agency in Polybius' work.
