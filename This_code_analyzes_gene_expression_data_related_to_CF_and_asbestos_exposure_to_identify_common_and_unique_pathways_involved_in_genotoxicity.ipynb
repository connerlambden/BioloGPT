### Gene Expression Analysis of CF and Asbestos Exposure

```python
import pandas as pd

# Load datasets related to CF and asbestos exposure
data_cf = pd.read_csv('cf_gene_expression.csv')
data_asbestos = pd.read_csv('asbestos_gene_expression.csv')

# Analyze common pathways
common_genes = set(data_cf['gene']).intersection(set(data_asbestos['gene']))
print(f'Common genes involved in genotoxicity: {common_genes}')
```
