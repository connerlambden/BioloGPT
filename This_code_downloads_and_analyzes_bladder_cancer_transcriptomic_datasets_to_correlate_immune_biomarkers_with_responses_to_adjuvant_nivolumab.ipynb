This notebook loads bladder cancer gene expression data to explore correlations between PD-L1 levels and immune scores.

```python
import pandas as pd
url = 'https://www.ebi.ac.uk/arrayexpress/experiments/E-GEOD-30522'
df = pd.read_csv(url, sep='\t')
print(df.head())
```

We then compute the Pearson correlation between PD-L1 expression and an immune response score.

```python
from scipy.stats import pearsonr
correlation, pvalue = pearsonr(df['PDL1'], df['Immune_Score'])
print(f'Correlation: {correlation}, p-value: {pvalue}')
```

Finally, we visualize the relationship using Plotly.

```python
import plotly.express as px
fig = px.scatter(df, x='PDL1', y='Immune_Score', title='PDL1 vs Immune Score')
fig.show()
```
