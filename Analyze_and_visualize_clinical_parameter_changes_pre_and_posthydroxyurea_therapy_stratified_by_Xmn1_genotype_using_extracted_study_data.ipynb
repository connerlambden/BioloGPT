This notebook downloads the extracted parameter data (ATR, Height, HbF) and compares pre and post-treatment values for CT and CC genotypes.

```python
import pandas as pd
import plotly.express as px

data = {
    'Parameter': ['ATR', 'Height', 'HbF']*2,
    'Before': [7.13, 112.66, 24.67, 7.5, 108.29, 26.5],
    'After': [3.36, 115.86, 31.81, 5.0, 116.59, 35.38],
    'Genotype': ['CT']*3 + ['CC']*3
}
df = pd.DataFrame(data)
fig = px.bar(df, x='Parameter', y=['Before','After'], barmode='group', facet_col='Genotype', title='Therapy Outcomes by Genotype')
fig.show()
```

The above code creates grouped bar plots to visualize changes in clinical parameters for each genotype.
