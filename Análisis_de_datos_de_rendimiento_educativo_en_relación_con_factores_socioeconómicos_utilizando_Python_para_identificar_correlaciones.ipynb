Análisis de datos de rendimiento educativo en relación con factores socioeconómicos.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar datos de rendimiento y factores socioeconómicos
data = pd.read_csv('pisa_data.csv')

# Visualizar correlaciones
sns.heatmap(data.corr(), annot=True)
plt.title('Correlaciones entre rendimiento y factores socioeconómicos')
plt.show()
```

Este análisis permite identificar qué factores socioeconómicos tienen mayor impacto en el rendimiento académico.

```python
# Análisis de regresión para predecir rendimiento basado en factores socioeconómicos
import statsmodels.api as sm
X = data[['factor1', 'factor2', 'factor3']]
y = data['rendimiento']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
model.summary()
```
