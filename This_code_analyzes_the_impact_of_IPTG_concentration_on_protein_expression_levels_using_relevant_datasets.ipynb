This notebook will analyze the relationship between IPTG concentration and protein yield based on experimental data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('ipgt_expression_data.csv')

# Plotting IPTG concentration vs protein yield
plt.figure(figsize=(10,6))
plt.scatter(data['IPTG_concentration'], data['protein_yield'])
plt.title('IPTG Concentration vs Protein Yield')
plt.xlabel('IPTG Concentration (mM)')
plt.ylabel('Protein Yield (mg/L)')
plt.grid()
plt.show()
```

The analysis shows how varying IPTG concentrations affect protein yield, helping to optimize expression conditions.
