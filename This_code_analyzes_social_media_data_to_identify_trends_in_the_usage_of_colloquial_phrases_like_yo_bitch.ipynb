### Data Collection
Collect social media posts containing the phrase 'yo bitch' to analyze its usage patterns.

```python
import pandas as pd

# Load dataset
posts = pd.read_csv('social_media_data.csv')

# Filter posts containing the phrase
filtered_posts = posts[posts['content'].str.contains('yo bitch')]

# Analyze frequency
frequency = filtered_posts['content'].value_counts()
```
