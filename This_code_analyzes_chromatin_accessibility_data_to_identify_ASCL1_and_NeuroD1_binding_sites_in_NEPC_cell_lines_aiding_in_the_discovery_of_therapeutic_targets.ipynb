This notebook will analyze chromatin accessibility data from NEPC cell lines to identify potential ASCL1 and NeuroD1 binding sites.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load chromatin accessibility data
data = pd.read_csv('chromatin_accessibility_data.csv')

# Analyze binding sites
binding_sites = data[data['factor'].isin(['ASCL1', 'NeuroD1'])]

# Visualize results
plt.figure(figsize=(10,6))
plt.hist(binding_sites['accessibility'], bins=30, color='purple', alpha=0.7)
plt.title('Chromatin Accessibility at ASCL1 and NeuroD1 Binding Sites')
plt.xlabel('Accessibility Level')
plt.ylabel('Frequency')
plt.show()
```

The analysis reveals the distribution of chromatin accessibility levels at ASCL1 and NeuroD1 binding sites, providing insights into their regulatory roles.
