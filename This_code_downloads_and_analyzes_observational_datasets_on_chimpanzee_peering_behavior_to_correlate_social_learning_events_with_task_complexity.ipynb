Import and preprocess the dataset containing peering events, role model diversity, and task complexity metrics.

```python
import pandas as pd

data = pd.read_csv('chimp_peering_data.csv')
print(data.head())
```

Perform statistical analysis to evaluate the relationship between social tolerance indices and tool use efficiency.

```python
import statsmodels.api as sm

X = data[['social_tolerance', 'role_model_diversity']]
Y = data['tool_use_efficiency']
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
print(model.summary())
```

Visualize the relationship using Plotly for an interactive graph.

```python
import plotly.express as px

fig = px.scatter(data, x='role_model_diversity', y='tool_use_efficiency', color='social_tolerance', title='Role Model Diversity vs Tool Use Efficiency')
fig.show()
```
