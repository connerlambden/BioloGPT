Download and preprocess RNA-seq data (GEO: GSE283266) from THAP1-depleted cells to determine differential expression levels of proteasome subunits, with a focus on PSMB5.

```python
import pandas as pd
import numpy as np
# Load count matrix for THAP1 knockout vs control
counts = pd.read_csv('GSE283266_counts.csv', index_col=0)
# Filter for proteasome genes
proteasome_genes = ['PSMB5', 'PSMB1', 'PSMB2']
df_proteasome = counts.loc[counts.index.intersection(proteasome_genes)]
# Output differential expression for PSMB5
print('PSMB5 expression levels:', df_proteasome.loc['PSMB5'])
# Additional analysis would follow using DESeq2 in R or similar tools in Python
```
