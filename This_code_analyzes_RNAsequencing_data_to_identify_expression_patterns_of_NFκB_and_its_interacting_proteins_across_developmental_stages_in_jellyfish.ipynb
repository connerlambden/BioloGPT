### RNA-sequencing Data Analysis for NF-κB Expression in Jellyfish

```python
import pandas as pd

# Load RNA-seq data
rna_data = pd.read_csv('path_to_rna_seq_data.csv')

# Analyze expression levels of NF-κB and interacting proteins
expression_summary = rna_data.groupby('gene').mean()
expression_summary
```

### Summary of Expression Levels
This analysis provides insights into the expression levels of NF-κB and its interacting proteins across different developmental stages.

```python
# Visualize expression levels
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(expression_summary.index, expression_summary['expression_level'])
plt.title('Expression Levels of NF-κB and Interacting Proteins')
plt.xlabel('Gene')
plt.ylabel('Expression Level')
plt.xticks(rotation=45)
plt.show()
```
