### GDF-15 Expression Analysis in Atherosclerosis

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load GDF-15 expression data
expression_data = pd.read_csv('gdf15_expression.csv')

# Analyze correlation with atherosclerosis markers
correlation = expression_data.corr()

# Plotting the correlation matrix
plt.figure(figsize=(10, 8))
plt.title('Correlation of GDF-15 with Atherosclerosis Markers')
plt.imshow(correlation, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.show()
```
