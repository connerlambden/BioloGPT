### RNA-seq Analysis of PGR Mutants Under Hypoxia

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load RNA-seq data
wild_type_data = pd.read_csv('wild_type_hypoxia.csv')
pgr_mutant_data = pd.read_csv('pgr_mutant_hypoxia.csv')

# Perform differential expression analysis
results = stats.ttest_ind(wild_type_data['expression'], pgr_mutant_data['expression'])

# Compile results
results_df = pd.DataFrame({'gene': wild_type_data['gene'], 'p_value': results.pvalue})
results_df['significant'] = results_df['p_value'] < 0.05
results_df.to_csv('differential_expression_results.csv', index=False)
```
