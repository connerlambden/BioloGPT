### Analysis of Protein Molecular Weight Distribution Using Sephadex Gel Filtration

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for molecular weights and elution volumes
data = {'Protein': ['BSA tetramer', 'BSA trimer', 'BSA dimer', 'BSA monomer'],
        'Molecular Weight (kDa)': [268, 201, 143, 67],
        'Elution Volume (mL)': [64.5, 58.6, 51.2, 40.7]}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['Protein'], df['Molecular Weight (kDa)'], color='skyblue')
plt.title('Molecular Weight Distribution of Proteins in Sephadex Gel Filtration')
plt.xlabel('Protein Type')
plt.ylabel('Molecular Weight (kDa)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

### Discussion
This analysis provides a visual representation of the molecular weight distribution of proteins separated by Sephadex, highlighting the effectiveness of gel filtration in protein purification.
