# Force Analysis of Molecular Motors
This notebook analyzes the force exerted by molecular motors under varying conditions.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load force data
force_data = pd.read_csv('force_data.csv')

# Analyze and plot the data
plt.figure(figsize=(10,6))
plt.bar(force_data['Force'], force_data['Count'])
plt.title('Force Distribution of Molecular Motors')
plt.xlabel('Force (pN)')
plt.ylabel('Count')
plt.show()
```
