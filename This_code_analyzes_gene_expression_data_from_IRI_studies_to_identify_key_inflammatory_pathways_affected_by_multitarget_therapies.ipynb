### Gene Expression Analysis in IRI Studies
This analysis focuses on gene expression changes in IRI models, comparing multi-target and single-target therapies.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load gene expression data
data = pd.read_csv('gene_expression_iri.csv')

# Analyze and visualize the data
plt.figure(figsize=(10,6))
plt.bar(data['Pathway'], data['Expression_Change'])
plt.title('Gene Expression Changes in IRI')
plt.xlabel('Inflammatory Pathways')
plt.ylabel('Expression Change')
plt.xticks(rotation=45)
plt.show()
```
