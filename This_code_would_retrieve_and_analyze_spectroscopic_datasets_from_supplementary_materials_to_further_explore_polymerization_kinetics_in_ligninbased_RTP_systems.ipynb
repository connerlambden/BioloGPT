Step-by-step extraction and analysis of NMR and FT-IR datasets from the supplementary information.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Code to load the NMR dataset and plot spectral changes pre- and post-UV irradiation
nmr_data = pd.read_csv('nmr_data.csv')
plt.figure(figsize=(8,4))
plt.plot(nmr_data['ppm'], nmr_data['intensity'], label='Before UV')
plt.plot(nmr_data['ppm'], nmr_data['intensity_uv'], label='After UV')
plt.xlabel('Chemical Shift (ppm)')
plt.ylabel('Intensity')
plt.title('NMR Spectral Comparison')
plt.legend()
plt.show()
```

This visualization aids in confirming the decrease in double bond signals indicative of polymerization.

```python
# Additional code for FT-IR data analysis would follow a similar structure
import numpy as np
ftir_data = pd.read_csv('ftir_data.csv')
plt.figure(figsize=(8,4))
plt.plot(ftir_data['wavenumber'], ftir_data['absorbance'], label='Pre-reaction')
plt.plot(ftir_data['wavenumber'], ftir_data['absorbance_post'], label='Post-reaction')
plt.xlabel('Wavenumber (cm-1)')
plt.ylabel('Absorbance')
plt.title('FT-IR Transformation')
plt.legend()
plt.show()
```
