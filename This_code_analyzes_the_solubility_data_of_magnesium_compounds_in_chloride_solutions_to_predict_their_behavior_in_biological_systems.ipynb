### Solubility Analysis of Magnesium Compounds in Chloride Solutions

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for solubility of Mg(OH)2 and MgCl2 in chloride solutions
data = {'Concentration (M)': [0, 0.1, 0.2, 0.3, 0.4],
        'Mg(OH)2 Solubility (g/L)': [0.1, 0.15, 0.2, 0.25, 0.3],
        'MgCl2 Solubility (g/L)': [0.5, 0.6, 0.7, 0.8, 0.9]}

# Create DataFrame
solubility_df = pd.DataFrame(data)

# Plotting the solubility data
plt.figure(figsize=(10, 6))
plt.plot(solubility_df['Concentration (M)'], solubility_df['Mg(OH)2 Solubility (g/L)'], label='Mg(OH)2', marker='o')
plt.plot(solubility_df['Concentration (M)'], solubility_df['MgCl2 Solubility (g/L)'], label='MgCl2', marker='o')
plt.title('Solubility of Magnesium Compounds in Chloride Solutions')
plt.xlabel('Chloride Concentration (M)')
plt.ylabel('Solubility (g/L)')
plt.legend()
plt.grid()
plt.show()
```

This analysis provides insights into how chloride concentration affects the solubility of magnesium compounds, which is crucial for their application in biological systems.
