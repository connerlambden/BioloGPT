### Genetic Variant Analysis for XP

```python
import pandas as pd

# Load genetic variant data
variant_data = pd.read_csv('xp_variants.csv')

# Analyze the frequency of mutations
mutation_counts = variant_data['mutation'].value_counts()
mutation_counts
```

### Summary of Mutation Frequencies

```python
mutation_counts.plot(kind='bar', title='Mutation Frequencies in XP Patients')
```
