This notebook will analyze RNA-seq data from HD models to identify changes in synaptic protein expression.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load RNA-seq data
rna_seq_data = pd.read_csv('rna_seq_data.csv')

# Perform differential expression analysis
results = stats.ttest_ind(rna_seq_data['control'], rna_seq_data['hd_model'])

# Output results
results_df = pd.DataFrame({'Gene': rna_seq_data['Gene'], 'p-value': results.pvalue})
results_df.to_csv('differential_expression_results.csv', index=False)
```
