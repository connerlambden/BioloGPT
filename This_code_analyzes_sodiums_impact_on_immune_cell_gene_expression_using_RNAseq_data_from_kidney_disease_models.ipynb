### Analyzing Sodium's Impact on Immune Cells
This notebook will explore gene expression changes in immune cells exposed to varying sodium levels.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load RNA-seq data
rna_data = pd.read_csv('sodium_immune_cells.csv')

# Visualize gene expression changes
sns.boxplot(x='sodium_level', y='gene_expression', data=rna_data)
plt.title('Gene Expression in Immune Cells at Different Sodium Levels')
plt.show()
```
