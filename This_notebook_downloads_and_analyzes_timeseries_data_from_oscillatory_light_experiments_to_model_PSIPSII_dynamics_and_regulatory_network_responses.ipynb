```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Placeholder for downloading dataset
# data = pd.read_csv('pea_oscillating_light_timeseries.csv')

time = np.array([0, 60, 120, 180, 240, 300])
psi = np.array([100, 90, 80, 78, 80, 90])
psii = np.array([100, 85, 75, 70, 75, 85])

plt.figure(figsize=(8,5))
plt.plot(time, psi, '-o', label='PSI Open Centers (%)', color='#FF9933')
plt.plot(time, psii, '-o', label='PSII Efficiency (%)', color='#6A0C76')
plt.xlabel('Time (s)')
plt.ylabel('Relative Activity (%)')
plt.title('PSI and PSII Dynamics Under Oscillating Light')
plt.legend()
plt.show()
```
