### Genetic Diversity Analysis of Thorny Skate Population

```python
import pandas as pd
import numpy as np

# Load genetic data
data = pd.read_csv('thorny_skate_genetic_data.csv')

# Calculate genetic diversity metrics
heterozygosity = data['heterozygosity'].mean()

# Output results
print('Average Heterozygosity:', heterozygosity)
```
