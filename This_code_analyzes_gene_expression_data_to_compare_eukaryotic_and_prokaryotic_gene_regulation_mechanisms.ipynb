# Gene Expression Analysis of Eukaryotic vs Prokaryotic Cells

```python
import pandas as pd

eukaryotic_data = pd.read_csv('eukaryotic_gene_expression.csv')
prokaryotic_data = pd.read_csv('prokaryotic_gene_expression.csv')

# Analyze differences in gene expression
comparison = eukaryotic_data.mean() - prokaryotic_data.mean()
comparison.to_csv('gene_expression_comparison.csv')
```

This analysis compares average gene expression levels between eukaryotic and prokaryotic cells, highlighting differences in regulatory mechanisms.

```python
# Visualize the comparison
import matplotlib.pyplot as plt

plt.bar(comparison.index, comparison.values)
plt.title('Gene Expression Comparison')
plt.xlabel('Genes')
plt.ylabel('Expression Difference')
plt.show()
```
