### Gene Expression Analysis of Mid-Insula and BNST
This analysis will focus on identifying differentially expressed genes in the mid-insula and BNST under stress conditions.

```python
import pandas as pd
import numpy as np

# Load gene expression data
gene_data = pd.read_csv('gene_expression_data.csv')

# Identify differentially expressed genes
differential_genes = gene_data[gene_data['p_value'] < 0.05]

# Output results
differential_genes.to_csv('differential_genes.csv', index=False)
```
