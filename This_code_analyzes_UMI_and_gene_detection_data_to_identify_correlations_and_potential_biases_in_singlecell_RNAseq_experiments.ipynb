### Analyzing UMI and Gene Detection Data
This analysis will explore the relationship between UMI counts and detected genes across different experiments.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for UMI and gene counts
data = {
    'Experiment': ['Exp1', 'Exp2'],
    'Median_UMI_Counts': [1194, 1954],
    'Median_Genes_Detected': [773, 1156]
}
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.bar(df['Experiment'], df['Median_UMI_Counts'], color='blue', alpha=0.6, label='Median UMI Counts')
plt.bar(df['Experiment'], df['Median_Genes_Detected'], color='orange', alpha=0.6, label='Median Genes Detected')
plt.ylabel('Counts')
plt.title('Comparison of UMI Counts and Detected Genes')
plt.legend()
plt.show()
```

This plot visualizes the relationship between UMI counts and detected genes, highlighting the discrepancies observed in the experiments.
