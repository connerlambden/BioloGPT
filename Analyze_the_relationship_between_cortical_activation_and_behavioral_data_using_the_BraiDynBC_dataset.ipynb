This notebook will analyze the BraiDyn-BC dataset to explore the relationship between cortical activation and behavioral changes during operant tasks.

```python
import pandas as pd

# Load the BraiDyn-BC dataset
dataset = pd.read_csv('braidyn_bc_data.csv')

# Display the first few rows of the dataset
dataset.head()
```

The dataset contains information on cortical activity and behavioral responses during operant tasks.

```python
# Analyze the correlation between cortical activation and behavioral performance
correlation = dataset.corr()
correlation
```

This analysis will help identify significant relationships between neural activity and behavior.

```python
# Visualize the correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Cortical Activation and Behavioral Correlation')
plt.show()
```
