### Analysis of Royal Jelly Effects on Sleep and Weight in Mice
This notebook will analyze the effects of royal jelly supplementation on sleep stability and body weight in diet-induced obese mice using the provided datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from the study
data = {
    'Group': ['ND', 'ND + RJ', 'HFD', 'HFD + RJ'],
    'Body Weight (g)': [29.1, 28.2, 41.4, 34.5],
    'NREM Sleep Stability': [1.0, 1.1, 0.8, 1.2]  # Hypothetical values for illustration
}

df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.bar(df['Group'], df['Body Weight (g)'], color='blue', alpha=0.6, label='Body Weight (g)')
plt.ylabel('Body Weight (g)')
plt.title('Impact of Royal Jelly on Body Weight in Mice')
plt.legend()
plt.show()
```
