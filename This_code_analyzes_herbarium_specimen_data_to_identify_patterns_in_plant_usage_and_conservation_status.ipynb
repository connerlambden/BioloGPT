### Herbarium Data Analysis
This notebook analyzes the collected herbarium data to identify trends in plant usage and conservation.

```python
import pandas as pd

# Load herbarium data
data = pd.read_csv('herbarium_data.csv')

# Analyze usage patterns
usage_patterns = data['usage'].value_counts()
usage_patterns
```

### Results
The analysis reveals the most commonly used plants and their conservation status.

```python
# Display results
usage_patterns.plot(kind='bar', title='Plant Usage Patterns')
```
