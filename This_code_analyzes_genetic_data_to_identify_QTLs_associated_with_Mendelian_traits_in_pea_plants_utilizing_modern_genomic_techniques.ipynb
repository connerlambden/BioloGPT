This notebook will guide you through the process of analyzing genetic data from pea plants to identify QTLs.

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load genetic data
data = pd.read_csv('pea_genetic_data.csv')

# Analyze QTLs
qtl_analysis = data.groupby('trait').agg({'marker': 'count', 'phenotype': 'mean'})
qtl_analysis.to_csv('qtl_analysis_results.csv')
```

The results of the QTL analysis will provide insights into the genetic basis of traits studied by Mendel.

```python
# Display results
print(qtl_analysis)
```
