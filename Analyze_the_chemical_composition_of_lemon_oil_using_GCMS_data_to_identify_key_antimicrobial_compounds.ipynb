This notebook will analyze GC-MS data of lemon oil to identify and quantify its major components.

```python
import pandas as pd

data = pd.read_csv('lemon_oil_gc_ms.csv')
# Analyze the data to find major components and their concentrations.
```

The analysis will provide insights into the composition and potential health benefits of lemon oil.

```python
# Code to visualize the composition of lemon oil
import matplotlib.pyplot as plt
plt.bar(data['Component'], data['Percentage'])
plt.title('Composition of Lemon Oil')
plt.xlabel('Components')
plt.ylabel('Percentage (%)')
plt.show()
```
