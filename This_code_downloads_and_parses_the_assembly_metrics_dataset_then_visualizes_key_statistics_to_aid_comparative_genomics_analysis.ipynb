This section downloads the relevant genome assembly metrics and sets up the visualization for comparing core statistics such as genome size, N50 metrics, and gene counts.

```python
import pandas as pd
import plotly.express as px

# Define the dataset
data = {
    'Metric': ['Genome Size (Mb)', 'Contig N50 (Kb)', 'Scaffold N50 (Mb)', 'Protein Coding Genes'],
    'Value': [516.11, 491.30, 16.24, 17739]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plot the metrics
fig = px.bar(df, x='Metric', y='Value', title='Key Assembly Metrics for Saccharina japonica', color_discrete_sequence=['#6A0C76'])
fig.show()
```
