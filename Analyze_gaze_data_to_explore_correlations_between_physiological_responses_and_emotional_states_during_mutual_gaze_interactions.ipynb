This notebook analyzes gaze data to understand the relationship between physiological responses and emotional states.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gaze_data.csv')
sns.scatterplot(data=data, x='heart_rate_variability', y='cone_of_direct_gaze')
plt.title('Correlation between HRV and Cone of Direct Gaze')
plt.show()
```

The scatter plot visualizes the correlation between heart rate variability and the cone of direct gaze.
