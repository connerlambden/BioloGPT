This section downloads and processes single-cell RNA-seq data from GBM samples to evaluate gene expression changes after oncolytic virus treatment.

```python
import scanpy as sc
adata = sc.read('path_to_gbm_scRNAseq_data.h5ad')
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.tl.rank_genes_groups(adata, groupby='treatment', method='wilcoxon')
sc.pl.rank_genes_groups(adata, n_genes=20, sharey=False)
```

The code above performs normalization, log-transformation, and differential gene expression analysis to pinpoint pathways involved in both cytolysis and immune activation.

```python
# Additional visualization using Plotly
import plotly.express as px
fig = px.bar(x=adata.uns['rank_genes_groups']['names']['treated'][:10], y=adata.uns['rank_genes_groups']['scores']['treated'][:10], labels={'x':'Gene', 'y':'Score'}, title='Top Differentially Expressed Genes')
fig.show()
```
