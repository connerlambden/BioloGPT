We begin by loading the microbial CUE and heterotrophic respiration datasets and examine the temperature-CUE relationship.

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('https://github.com/KailiangYu/Microbial-carbon-useefficiency/raw/main/data.csv')
fig = px.scatter(df, x='MAT', y='CUE', trendline='lowess', title='Nonlinear Response of Microbial CUE')
fig.show()
```

The notebook then models the SOC response using segmented regression analysis to capture the threshold behavior.

```python
import statsmodels.api as sm
import numpy as np

# Example code for segmented regression
x = df['MAT']
y = df['CUE']

# Fit a simple linear model as a starting point
model = sm.OLS(y, sm.add_constant(x)).fit()
print(model.summary())
```
