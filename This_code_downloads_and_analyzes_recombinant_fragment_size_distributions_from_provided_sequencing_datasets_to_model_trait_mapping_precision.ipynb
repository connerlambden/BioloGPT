In this notebook, we load the sequencing data from the recombinant libraries, compute the distribution of recombination fragment sizes, and visualize these using Plotly to better understand the patterning across different E. coli strains.

```python
import pandas as pd
import numpy as np
import plotly.express as px

# Load dataset from provided DOI data link (assuming data is available as CSV)
data = pd.read_csv('path_to_recombinant_fragments.csv')

# Compute fragment size distribution
fragment_sizes = data['fragment_size']

df = pd.DataFrame({'Fragment Size (kb)': fragment_sizes / 1000})

# Plot distribution
fig = px.histogram(df, x='Fragment Size (kb)', nbins=50, title='Recombination Fragment Size Distribution')
fig.update_layout(title_font_color='#6A0C76', xaxis_title='Fragment Size (kb)', yaxis_title='Frequency')
fig.show()
```

The above code calculates and plots the fragment size distribution, which directly informs how fragment size variability can enable precise trait mapping.
