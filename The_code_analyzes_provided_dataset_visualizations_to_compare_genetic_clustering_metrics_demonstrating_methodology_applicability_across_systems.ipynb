This notebook demonstrates data loading and visualization of genetic variability metrics. The dataset, extracted from Cannabis sativa studies, serves as a proxy for clustering analyses relevant to bacterial genetic structure.

```python
import pandas as pd
import plotly.express as px

data = pd.DataFrame({
    'Sample': ['Sample 1 (Wisconsin)', 'Sample 2 (North Dakota)'],
    'THC': [0.04, 0.07]
})

fig = px.bar(data, x='Sample', y='THC', title='THC Content in Feral Cannabis Samples', labels={'THC': 'THC (%)'})
fig.show()
```

The above steps detail data loading and simple bar-plot creation using Plotly. This approach can be adapted to bacterial genetic datasets to visualize clustering and drift metrics.
