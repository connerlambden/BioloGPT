This notebook downloads the genetic marker dataset of S. halieti and performs analysis of sequence similarity percentages using available accession numbers.

```python
import pandas as pd

data = {'Marker': ['18S rRNA', '28S rRNA', 'cox1', 'ITS1'],
        'Accession': ['PQ270244', 'PQ270246', 'PQ274874', 'PQ276104'],
        'Similarity': ['99.9-100%', '99.9-100%', '98.4-100%', '96.3%']}

df = pd.DataFrame(data)

df
```

Next, the notebook performs visualization of the genetic similarity data to verify genetic identity confirmation.

```python
import plotly.express as px

fig = px.bar(df, x='Marker', y=[90, 95, 98, 96], text='Similarity', title='Genetic Similarity of S. halieti Markers')
fig.update_layout(title_font_color='#6A0C76')
fig.show()
```
