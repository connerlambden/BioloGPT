This notebook will analyze mutational data from directed evolution experiments to identify instances of multiple mutations at the same site.

```python
import pandas as pd

# Load mutational data
data = pd.read_csv('mutational_data.csv')

# Analyze occurrences of multiple mutations at the same site
multiple_mutations = data[data.duplicated(['site'], keep=False)]

# Display results
multiple_mutations
```

The analysis reveals how often specific sites in proteins are subject to multiple mutations, providing insights into the adaptive walk process.

```python
# Display the results in a user-friendly format
multiple_mutations_summary = multiple_mutations.groupby('site').size().reset_index(name='count')
print(multiple_mutations_summary)
```
