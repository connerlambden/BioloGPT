### Data Analysis of Biocrude Properties
This section will analyze the physicochemical properties of biocrude obtained from different HTL temperatures.

```python
import pandas as pd

# Load the dataset containing biocrude properties
dataset = pd.read_csv('biocrude_properties.csv')

# Display the first few rows of the dataset
dataset.head()
```

### Statistical Analysis
Perform statistical analysis to determine the optimal HTL temperature for biocrude production.

```python
# Statistical analysis
summary_stats = dataset.describe()
summary_stats
```

### Visualization
Visualize the relationship between HTL temperature and the heating value of biocrude.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(dataset['HTL_Temperature'], dataset['Heating_Value'], color='blue')
plt.title('Heating Value vs HTL Temperature')
plt.xlabel('HTL Temperature (°C)')
plt.ylabel('Heating Value (MJ/kg)')
plt.grid()
plt.show()
```
