# Analysis of Anti-Proliferative Activity Data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for analysis
data = {'Compound': ['1h', '1c'], 'IC50': [224.14, 210.03]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Compound'], df['IC50'], color='blue')
plt.title('Anti-Proliferative Activity of Hydroxy Imidazole N-Oxides')
plt.xlabel('Compounds')
plt.ylabel('IC50 (mg/ml)')
plt.show()
```

This analysis provides a visual representation of the anti-proliferative activities of the synthesized compounds.
