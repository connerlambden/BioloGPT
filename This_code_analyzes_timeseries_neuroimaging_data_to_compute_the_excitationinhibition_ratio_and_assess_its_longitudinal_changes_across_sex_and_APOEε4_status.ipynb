The notebook downloads and preprocesses resting-state fMRI and DTI data from the provided OASIS repository to simulate E-I balance trajectories.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
# Load neuroimaging data and demographic info
data = pd.read_csv('https://www.oasis-brains.org/data_sample.csv')
# Compute excitation-inhibition ratio for each subject over time
model = sm.MixedLM(data['EIR'], data[['time', 'APOE_Status', 'Sex']], groups=data['SubjectID'])
result = model.fit()
print(result.summary())
```

The above code fits a linear mixed model to the data, capturing the interaction effects among time, gender, and APOE-ε4 status.

```python
# Plotting the E-I trajectories
import plotly.express as px
fig = px.line(data, x='time', y='EIR', color='APOE_Status', facet_col='Sex', title='E-I Balance Trajectories')
fig.show()
```

This visualization aids in revealing the sex-specific hyperexcitable patterns in APOE-ε4 carriers over time.
