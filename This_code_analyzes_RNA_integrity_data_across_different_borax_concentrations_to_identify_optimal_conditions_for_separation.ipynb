### RNA Integrity Analysis with Borax Concentration Data

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Concentration (μM)': [50, 100, 200, 300, 500, 1000, 2000, 5000, 10000],
        'RNA Integrity (%)': [100, 95, 90, 85, 70, 50, 30, 10, 0]}
df = pd.DataFrame(data)
plt.plot(df['Concentration (μM)'], df['RNA Integrity (%)'], marker='o')
plt.xscale('log')
plt.title('Effect of Borax Concentration on RNA Integrity')
plt.xlabel('Borax Concentration (μM)')
plt.ylabel('RNA Integrity (%)')
plt.grid()
plt.show()
```

### This analysis visualizes the relationship between borax concentration and RNA integrity, helping to identify optimal conditions for RNA separation.
