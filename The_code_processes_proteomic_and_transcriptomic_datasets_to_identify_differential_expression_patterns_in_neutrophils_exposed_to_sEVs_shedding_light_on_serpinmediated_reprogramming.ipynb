Retrieve and integrate proteomic data from ProteomeXchange (PXD057119) and RNA-seq datasets from neutrophils treated with serpin-rich versus serpin-depleted sEVs.

```python
import pandas as pd
proteomics = pd.read_csv('PXD057119_proteomics.csv')
rna_seq = pd.read_csv('RNAseq_neutrophils.csv')
merged_data = pd.merge(proteomics, rna_seq, on='gene_id')
print(merged_data.head())
```
