This notebook retrieves the irradiation dose-response data and performs statistical plotting to correlate doses with cell viability and antigen yield from the paper.

```python
import pandas as pd
import plotly.express as px

data = {'Dose (Gy)': [5, 10, 15, 20, 25],
        'Cell Viability (%)': [90, 70, 50, 30, 10],
        '146S Antigen Level': [1, 2, 5, 4, 3]}
df = pd.DataFrame(data)

fig1 = px.line(df, x='Dose (Gy)', y='Cell Viability (%)', markers=True, title='Cell Viability vs Dose')
fig2 = px.line(df, x='Dose (Gy)', y='146S Antigen Level', markers=True, title='Antigen Level vs Dose')

fig1.show()
fig2.show()
```
