### Data Analysis of CBP-O Binding Affinity

```python
import pandas as pd

# Load experimental data
cbp_data = pd.read_csv('cbp_binding_data.csv')

# Analyze binding affinities
binding_affinities = cbp_data.groupby('nucleotide')['affinity'].mean()
print(binding_affinities)
```
