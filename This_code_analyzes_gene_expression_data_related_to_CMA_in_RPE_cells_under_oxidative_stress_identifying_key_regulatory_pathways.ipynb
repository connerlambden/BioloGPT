### Gene Expression Analysis of CMA in RPE Cells
This notebook analyzes RNA-seq data to identify genes involved in chaperone-mediated autophagy under oxidative stress.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_data.csv')
sns.boxplot(x='condition', y='expression', data=data)
plt.title('Gene Expression Under Different Conditions')
plt.show()
```
