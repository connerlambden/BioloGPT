### Performance Analysis of Thermoelectric Sensor

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for the sensor performance
performance_data = {
    'Feature': ['Seebeck Coefficient (μV/°C)', 'Temperature Resolution (°C)', 'Gauge Factor', 'Stretchability (%)'],
    'Value': [37.33, 0.5, 1401.5, 45]
}

# Create DataFrame
df = pd.DataFrame(performance_data)

# Plotting the performance
plt.bar(df['Feature'], df['Value'], color='blue')
plt.title('Performance Metrics of Thermoelectric Sensor')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```
