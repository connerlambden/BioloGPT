Below is a step-by-step Jupyter Notebook code snippet for analyzing scRNA-seq data from GBM, focusing on identification of rare neoplastic-stemness cells.

```python
import scanpy as sc
import pandas as pd

# Load dataset (replace 'data_file.h5ad' with actual dataset)
adata = sc.read('data_file.h5ad')

# Preprocessing: filtering, normalization, and log transformation
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)

# Identify highly variable genes
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
adata = adata[:, adata.var['highly_variable']]

# PCA and neighborhood graph construction
sc.tl.pca(adata, svd_solver='arpack')
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)

# Clustering the cells
sc.tl.leiden(adata, resolution=0.5)

# UMAP visualization
sc.tl.umap(adata)
sc.pl.umap(adata, color=['leiden'], save='_clusters.png')

# Identify potential neoplastic-stemness cells based on marker gene expression (example markers)
markers = ['CD44', 'SOX2']
sc.pl.umap(adata, color=markers, save='_markers.png')

# Differential expression analysis
sc.tl.rank_genes_groups(adata, 'leiden', method='wilcoxon')
sc.pl.rank_genes_groups(adata, n_genes=20, save='_rank_genes.png')

# Save the processed data
adata.write('processed_gbm.h5ad')
```

The code above illustrates a complete workflow from data loading to clustering and marker gene visualization, important for identifying rare cell subpopulations such as the neoplastic-stemness cells described in the paper.
