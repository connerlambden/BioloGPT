### Data Analysis of Mental Disorders Among Palestinian Adolescents
This analysis focuses on the prevalence of mental disorders among Palestinian adolescents, utilizing data from the Galilee Study.

```python
import pandas as pd

# Load the dataset
url = 'https://example.com/galilee_study_data.csv'
data = pd.read_csv(url)

# Analyze the prevalence of disorders
prevalence = data.groupby('ethnicity')['disorder'].value_counts(normalize=True) * 100
print(prevalence)
```
