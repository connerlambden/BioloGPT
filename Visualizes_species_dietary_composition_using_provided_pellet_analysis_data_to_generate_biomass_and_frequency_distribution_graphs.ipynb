This notebook loads the dietary dataset and plots biomass percentage and prey frequency distributions using Plotly.

```python
import plotly.express as px
import pandas as pd

data = pd.DataFrame([
    {'species': 'Didelphis pernigra', 'average_mass': 400, 'number_individuals': 22, 'percentage_frequency': 20.8, 'biomass_percentage': 66.18},
    {'species': 'Rattus rattus', 'average_mass': 162, 'number_individuals': 18, 'percentage_frequency': 17.0, 'biomass_percentage': 21.93},
    {'species': 'Reithrodontomys soederstroemi', 'average_mass': 15, 'number_individuals': 25, 'percentage_frequency': 23.6, 'biomass_percentage': 2.82},
    {'species': 'Eptesicus miradorensis', 'average_mass': 15.5, 'number_individuals': 2, 'percentage_frequency': 1.9, 'biomass_percentage': 0.23},
    {'species': 'Gastrotheca riobambae', 'average_mass': None, 'number_individuals': 1, 'percentage_frequency': None, 'biomass_percentage': 0}
])

fig1 = px.bar(data, x='species', y='biomass_percentage', title='Biomass Percentage by Prey Species', labels={'biomass_percentage':'Biomass (%)', 'species':'Prey Species'})
fig1.show()

fig2 = px.bar(data, x='species', y='percentage_frequency', title='Prey Frequency Percentage', labels={'percentage_frequency':'Frequency (%)', 'species':'Prey Species'})
fig2.show()
```

The resulting plots illustrate the relative contribution of each prey species in terms of biomass and occurrence, which is pivotal for interpreting the owl's dietary adaptations.
