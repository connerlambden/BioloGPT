### Data Analysis of Feeding Regimens on Zebrafish Behavior

```python
import pandas as pd

# Load the dataset
feeding_data = pd.read_csv('feeding_regimens_zebrafish.csv')

# Analyze the impact of feeding on swimming bouts
summary = feeding_data.groupby(['developmental_stage', 'feeding_regimen']).agg({'swimming_bouts': 'mean'}).reset_index()
print(summary)
```
