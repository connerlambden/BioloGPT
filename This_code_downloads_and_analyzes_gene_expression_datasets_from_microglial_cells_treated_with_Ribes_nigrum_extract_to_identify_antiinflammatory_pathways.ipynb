This notebook imports expression arrays from provided datasets and uses differential expression analysis to highlight key inflammatory genes.

```python
import pandas as pd
import numpy as np
# Load dataset using pandas
expression_data = pd.read_csv('path_to_microglial_expression_data.csv')
# Perform differential gene expression analysis
from scipy.stats import ttest_ind
# Assume two conditions: treated and control
treated = expression_data[expression_data['condition']=='treated']
control = expression_data[expression_data['condition']=='control']
results = {} 
for gene in expression_data.columns[2:]:
    stat, pval = ttest_ind(treated[gene], control[gene])
    results[gene] = pval
# Sort genes by p-value
sorted_genes = sorted(results.items(), key=lambda x: x[1])
print(sorted_genes[:10])
```

The above code evaluates expression differences between treated and control groups, highlighting potential anti-inflammatory targets modulated by the extract.
