Este análisis utilizará datos de expresión génica de vacuolas para identificar transportadores de fosfato en plantas.

```python
import pandas as pd

# Cargar datos de expresión génica
data = pd.read_csv('data_vacuoles.csv')

# Análisis de datos
data.describe()
```
