### Average Heterozygosity Calculation
This notebook calculates average heterozygosity (He) using allele frequency data from various populations.

```python
import numpy as np

# Example allele frequencies
allele_frequencies = np.array([0.3, 0.7])  # p and q for a biallelic locus

# Calculate average heterozygosity
He = 1 - np.sum(allele_frequencies**2)
He
```

### Results
The calculated average heterozygosity (He) provides a measure of genetic diversity for the given allele frequencies.

```python
# Display the result
print(f'Average Heterozygosity (He): {He:.4f}')
```
