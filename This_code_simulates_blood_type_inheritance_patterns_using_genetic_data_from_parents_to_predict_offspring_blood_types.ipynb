This notebook analyzes blood type inheritance using Punnett squares and genetic data.

```python
# Simulate blood type inheritance
# Define parent genotypes
jack_genotype = ['I_A', 'I_B']
jill_genotype = ['I_B', 'i']
# Calculate possible offspring genotypes
possible_offspring = [jack + jill for jack in jack_genotype for jill in jill_genotype]
```

The results show the potential blood types of the offspring.

```python
# Display results
set(possible_offspring)
```
