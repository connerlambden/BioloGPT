Este análisis utilizará datos de ADN antiguo para investigar patrones de migración y mestizaje en Europa.

```python
# Código para analizar datos de ADN antiguo
import pandas as pd
# Cargar datos
ancient_dna_data = pd.read_csv('ancient_dna.csv')
# Análisis de patrones de migración
migration_patterns = ancient_dna_data.groupby('population').mean()
```

Los resultados mostrarán cómo las poblaciones medievales se mezclaron y migraron a lo largo del tiempo.

```python
# Visualización de patrones de migración
import matplotlib.pyplot as plt
plt.bar(migration_patterns.index, migration_patterns['migration_rate'])
plt.title('Patrones de Migración en la Edad Media')
plt.show()
```
