Analyze gene expression data from IBD patients treated with MBD.

```python
import pandas as pd
# Load gene expression data
data = pd.read_csv('gene_expression_ibd.csv')
# Filter for JAK/STAT pathway genes
jak_stat_genes = data[data['gene'].isin(['JAK1', 'JAK2', 'STAT3', 'STAT5'])]
```

Perform statistical analysis to compare expression levels before and after MBD treatment.

```python
from scipy import stats
# Compare expression levels
before = jak_stat_genes['expression_before']
after = jak_stat_genes['expression_after']
t_stat, p_value = stats.ttest_ind(before, after)
```
