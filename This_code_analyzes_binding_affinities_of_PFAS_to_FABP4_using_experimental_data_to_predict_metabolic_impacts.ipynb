### Binding Affinity Analysis of PFAS to FABP4

```python
import pandas as pd

# Data on PFAS binding affinities
pfas_data = {'Compound': ['PFOA', 'PFDA', 'PFHxDA'], 'Kd (μM)': [0.5, 0.8, 0.6]}
df = pd.DataFrame(pfas_data)

# Display the DataFrame
df
```

### Analysis of Binding Affinities
