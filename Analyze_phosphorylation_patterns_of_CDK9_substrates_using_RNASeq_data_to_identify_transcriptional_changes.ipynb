This notebook will analyze RNA-Seq data to identify changes in gene expression associated with CDK9 phosphorylation.

```python
import pandas as pd
import numpy as np

# Load RNA-Seq data
rna_seq_data = pd.read_csv('path_to_rna_seq_data.csv')

# Analyze expression levels of CDK9 substrates
cdk9_substrates = ['SPT5', 'XRN2']
expression_levels = rna_seq_data[rna_seq_data['gene'].isin(cdk9_substrates)]

# Perform statistical analysis
results = expression_levels.groupby('condition').mean()
```

The analysis will provide insights into how CDK9 phosphorylation affects gene expression.

```python
# Visualize results
import matplotlib.pyplot as plt
results.plot(kind='bar')
plt.title('Expression Levels of CDK9 Substrates')
plt.xlabel('Condition')
plt.ylabel('Expression Level')
plt.show()
```
