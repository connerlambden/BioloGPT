### Binding Affinity Analysis of PET Tracers
This notebook will analyze the binding affinities of various PET tracers to protein fibrils using a cooperative model.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters for the model
ligand_concentration = np.linspace(0, 100, 100)
# Hypothetical binding affinity function
binding_affinity = ligand_concentration / (1 + ligand_concentration/10)

# Plotting the binding affinity
plt.plot(ligand_concentration, binding_affinity)
plt.title('Binding Affinity of PET Tracers')
plt.xlabel('Ligand Concentration (nM)')
plt.ylabel('Binding Affinity')
plt.grid(True)
plt.show()
```

### Discussion
This analysis provides insights into how ligand concentration affects binding affinity, illustrating the cooperative nature of binding.
