This notebook will analyze datasets comparing IgE levels from intranasal and intradermal allergen administration.

```python
import pandas as pd

# Load datasets
intranasal_data = pd.read_csv('intranasal_data.csv')
intradermal_data = pd.read_csv('intradermal_data.csv')

# Analyze IgE levels
combined_data = pd.concat([intranasal_data, intradermal_data])
results = combined_data.groupby('route')['IgE_level'].mean()
results.to_csv('IgE_comparison_results.csv')
```

The analysis will provide insights into the average IgE levels produced by each administration route.

```python
# Display results
print(results)
```
