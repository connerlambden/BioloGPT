### Nutrient Excretion Analysis in Guppies
This notebook analyzes the impact of predation risk on nutrient excretion in Trinidadian guppies.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Site': ['Guanapo', 'Aripo'], 'Mean Tissue N (%)': [9.1, 8.5], 'Mean Tissue P (%)': [3.2, 2.8]}
df = pd.DataFrame(data)

plt.bar(df['Site'], df['Mean Tissue N (%)'], color='blue', label='Mean Tissue N (%)')
plt.bar(df['Site'], df['Mean Tissue P (%)'], color='green', label='Mean Tissue P (%)', bottom=df['Mean Tissue N (%)'])
plt.title('Nutrient Excretion in Trinidadian Guppies')
plt.ylabel('Percentage (%)')
plt.legend()
plt.show()
```
