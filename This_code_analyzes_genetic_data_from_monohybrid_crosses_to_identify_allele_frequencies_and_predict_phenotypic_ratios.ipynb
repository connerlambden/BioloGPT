### Genetic Data Analysis for Monohybrid Crosses
This notebook will analyze genetic data from monohybrid crosses to determine allele frequencies and predict phenotypic ratios.

```python
import pandas as pd

def analyze_genetic_data(data):
    # Calculate allele frequencies
    total = data['count'].sum()
    data['frequency'] = data['count'] / total
    return data

# Example data
example_data = pd.DataFrame({'allele': ['P', 'p'], 'count': [75, 25]})
result = analyze_genetic_data(example_data)
print(result)
```
