```python
import pandas as pd
import plotly.express as px
# Load dataset for GIRK2 expression and ubiquitination levels
data = pd.read_csv('girK2_ubiquitination_data.csv')
fig = px.scatter(data, x='Ubiquitination_Level', y='MWM_Performance', color='Group', title='GIRK2 Ubiquitination vs Cognitive Performance')
fig.show()
```
