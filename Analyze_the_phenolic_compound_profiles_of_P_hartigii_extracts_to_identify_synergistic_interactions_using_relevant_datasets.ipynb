This notebook will analyze the phenolic profiles of P. hartigii extracts, focusing on identifying synergistic interactions among compounds.

```python
import pandas as pd

# Load phenolic compound data
phenolic_data = pd.read_csv('path_to_phenolic_data.csv')

# Analyze interactions
synergy_analysis = phenolic_data.groupby('Compound').agg({'Activity': 'sum'})
synergy_analysis.to_csv('synergy_analysis_results.csv')
```

The results of the synergy analysis will provide insights into which combinations of phenolic compounds yield the highest bioactivity.

```python
# Visualize results
import matplotlib.pyplot as plt

plt.bar(synergy_analysis.index, synergy_analysis['Activity'])
plt.title('Synergistic Activity of Phenolic Compounds')
plt.xlabel('Phenolic Compounds')
plt.ylabel('Activity')
plt.show()
```
