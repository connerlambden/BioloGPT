This notebook will simulate the dynamics of the beta hairpin loop in HIV capsid proteins to understand its role in stability.

```python
import mdtraj as md
# Load the HIV capsid structure
capsid = md.load('hiv_capsid.pdb')
# Run molecular dynamics simulation
md.run(capsid, time=100) # 100 ns simulation
```

The results will provide insights into the flexibility and interactions of the beta hairpin loop.

```python
# Analyze the trajectory
rmsd = md.rmsd(capsid, reference=capsid[0])
print(rmsd)
```
