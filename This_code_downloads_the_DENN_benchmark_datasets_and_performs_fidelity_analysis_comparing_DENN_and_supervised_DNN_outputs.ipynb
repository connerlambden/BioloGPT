Loading and exploring the DENN dataset to compare phase retrieval fidelity.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('https://zenodo.org/records/10938116')
plt.figure(figsize=(8,5))
plt.plot(data['supervised_fidelity'], label='Supervised DNN')
plt.plot(data['denn_fidelity'], label='DENN')
plt.xlabel('Test Samples')
plt.ylabel('Fidelity')
plt.title('Phase Retrieval Fidelity Comparison')
plt.legend()
plt.show()
```

This visualization highlights the fidelity improvements achieved by the DENN approach over traditional methods.
