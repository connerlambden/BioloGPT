### Data Analysis of Somatic Cell Counts and Microbial Composition

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for SCC
scc_data = {'Timepoint': ['T1', 'T2', 'T3'], 'SCC_TG': [1588000, 1588000, 148000], 'SCC_CG': [1629000, 1629000, 153000]}
scc_df = pd.DataFrame(scc_data)

# Plotting SCC
plt.figure(figsize=(10, 6))
plt.bar(scc_df['Timepoint'], scc_df['SCC_TG'], color='blue', label='Treatment Group (TG)')
plt.bar(scc_df['Timepoint'], scc_df['SCC_CG'], color='orange', label='Control Group (CG)', alpha=0.7)
plt.title('Somatic Cell Counts (SCC) Over Time')
plt.xlabel('Timepoints')
plt.ylabel('SCC (cells/mL)')
plt.legend()
plt.show()
```
