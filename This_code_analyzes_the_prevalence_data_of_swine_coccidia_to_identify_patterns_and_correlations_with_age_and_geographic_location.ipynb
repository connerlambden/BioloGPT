### Data Analysis of Swine Coccidia Prevalence
This notebook analyzes the prevalence of swine coccidia based on the data extracted from the study.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from the study
data = {
    'Species': ['Cystoisospora suis', 'Eimeria polita', 'Eimeria scabra'],
    'Prevalence (%)': [58.77, 3.31, 0.46]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['Species'], df['Prevalence (%)'], color='skyblue')
plt.title('Prevalence of Swine Coccidia Species')
plt.xlabel('Species')
plt.ylabel('Prevalence (%)')
plt.grid(axis='y')
plt.show()
```

### Discussion
This analysis provides insights into the distribution of coccidia species in swine populations, highlighting the dominance of <em>Cystoisospora suis</em>.
