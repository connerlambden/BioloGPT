# Analysis of OWCS Performance Data
This notebook analyzes the performance metrics of the Optical Wireless Communication System (OWCS) based on the provided data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for analysis
data = {
    'Thickness (mm)': [2, 5, 10],
    'Max Baud Rate (Mbit/s)': [6, 5.5, 5],
    'Optical Power (mW)': [55, 55, 55]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 5))
plt.bar(df['Thickness (mm)'], df['Max Baud Rate (Mbit/s)'], color='blue', label='Max Baud Rate')
plt.bar(df['Thickness (mm)'], df['Optical Power (mW)'], color='orange', label='Optical Power')
plt.xlabel('Tissue Thickness (mm)')
plt.ylabel('Value')
plt.title('OWCS Performance Metrics')
plt.legend()
plt.show()
```
