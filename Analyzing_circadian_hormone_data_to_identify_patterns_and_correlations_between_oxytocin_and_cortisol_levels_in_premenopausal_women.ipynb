This notebook will analyze circadian hormone data from pre-menopausal women, focusing on oxytocin and cortisol levels.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
oxytocin_data = pd.read_csv('oxytocin_levels.csv')
cortisol_data = pd.read_csv('cortisol_levels.csv')

# Merge datasets on time points
merged_data = pd.merge(oxytocin_data, cortisol_data, on='time')

# Plotting the data
plt.figure(figsize=(12, 6))
plt.plot(merged_data['time'], merged_data['oxytocin'], label='Oxytocin Levels')
plt.plot(merged_data['time'], merged_data['cortisol'], label='Cortisol Levels')
plt.title('Circadian Dynamics of Oxytocin and Cortisol')
plt.xlabel('Time of Day')
plt.ylabel('Hormone Levels')
plt.legend()
plt.show()
```

This analysis provides insights into the circadian dynamics of oxytocin and cortisol, highlighting potential interactions.
