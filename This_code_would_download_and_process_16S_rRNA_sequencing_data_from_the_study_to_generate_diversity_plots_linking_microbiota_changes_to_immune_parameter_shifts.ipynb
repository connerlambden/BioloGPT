Download and preprocess the 16S rRNA sequencing dataset, perform alpha and beta diversity analyses using Qiime2 outputs.

```python
import pandas as pd
import plotly.express as px
# Load study data
metadata = pd.read_csv('study_metadata.csv')
alpha_diversity = pd.read_csv('alpha_diversity.csv')
fig = px.line(alpha_diversity, x='Time_Point', y='Shannon_Index', markers=True, title='Alpha-Diversity Over Time')
fig.show()
```

This notebook section visualizes changes in microbiota diversity across time points, linking to systemic parameter shifts.

```python
# Further code could integrate immune marker data correlatively
import seaborn as sns
import matplotlib.pyplot as plt
merged_data = pd.merge(alpha_diversity, metadata, on='Participant_ID')
sns.lineplot(data=merged_data, x='Time_Point', y='Leukocyte_Count', hue='Participant_ID')
plt.title('Leukocyte Count Over Time')
plt.show()
```
