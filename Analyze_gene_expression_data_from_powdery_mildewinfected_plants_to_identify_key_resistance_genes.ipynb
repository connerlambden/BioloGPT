This notebook will analyze RNA-seq data to identify differentially expressed genes in response to powdery mildew infection.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load the dataset
data = pd.read_csv('powdery_mildew_expression_data.csv')

# Perform differential expression analysis
results = stats.ttest_ind(data['infected'], data['control'])

# Output results
results_df = pd.DataFrame({'Gene': data['Gene'], 'p-value': results.pvalue})
results_df.to_csv('differential_expression_results.csv', index=False)
```
