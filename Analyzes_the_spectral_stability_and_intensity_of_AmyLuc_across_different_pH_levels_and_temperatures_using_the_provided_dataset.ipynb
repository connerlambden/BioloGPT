## Analysis of AmyLuc Bioluminescence Properties

This notebook analyzes the stability and intensity of AmyLuc luciferase across various pH levels and temperatures using the provided experimental data.

```python
import pandas as pd
import plotly.express as px

data = pd.read_csv('bioluminescence_data.csv')
```

### Bioluminescence Intensity Comparison

The following graph compares the bioluminescence intensity of AmyLuc with CrtLuc and Mac-Luc.

```python
fig = px.bar(data, x='Luciferase', y='Bioluminescence_Intensity', color='Luciferase', title='Bioluminescence Intensity Comparison')
fig.show()
```

### Spectral Shift Analysis

This graph illustrates the spectral shifts observed in each luciferase at 37°C.

```python
fig = px.line(data, x='Luciferase', y='Spectral_Shift_nm', title='Spectral Shift at 37°C')
fig.show()
```

### Stability Assessment

Assessing the stability of each luciferase at physiological temperature.

```python
fig = px.scatter(data, x='Luciferase', y='Stability_at_37C', title='Stability of Luciferases at 37°C')
fig.show()
```
