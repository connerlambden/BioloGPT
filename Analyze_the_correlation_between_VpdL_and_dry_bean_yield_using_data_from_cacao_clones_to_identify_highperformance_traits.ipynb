### Data Analysis of VpdL and Yield in Cacao Clones
This notebook will analyze the relationship between VpdL and dry bean yield across different cacao clones.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
# Assuming a CSV file with columns 'Clone', 'VpdL', 'Yield'
data = pd.read_csv('cacao_data.csv')

# Visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='VpdL', y='Yield', hue='Clone')
plt.title('VpdL vs Dry Bean Yield in Cacao Clones')
plt.xlabel('VpdL (kPa)')
plt.ylabel('Yield (kg/ha)')
plt.show()
```
