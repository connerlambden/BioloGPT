This notebook imports the GH18 gene dataset from the paper, constructs a DataFrame, and visualizes gene molecular weights via a bar chart.

```python
import pandas as pd
import plotly.express as px

data = [
    {'gene_name': 'Cordyceps0G001980.1', 'number_of_amino_acids': 787, 'molecular_weight_kDa': 81.41, 'theoretical_pI': 7.92},
    {'gene_name': 'Cordyceps0G005200.1', 'number_of_amino_acids': 1782, 'molecular_weight_kDa': 19.91, 'theoretical_pI': 5.33},
    {'gene_name': 'Cordyceps0G005800.1', 'number_of_amino_acids': 1310, 'molecular_weight_kDa': 14.35, 'theoretical_pI': 5.55}
]
df = pd.DataFrame(data)
fig = px.bar(df, x='gene_name', y='molecular_weight_kDa', title='GH18 Gene Molecular Weights', labels={'molecular_weight_kDa': 'Molecular Weight (kDa)'}, color_discrete_sequence=['#6A0C76'])
fig.show()
```

This segment clearly outlines the dataset and the visualization approach. Expanding this analysis could involve multi-dimensional plotting with additional gene parameters.

```python
# Additional analysis code can involve correlation plots between molecular weight and theoretical pI
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x='molecular_weight_kDa', y='theoretical_pI', hue='gene_name', palette=['#6A0C76'])
plt.title('Correlation between Molecular Weight and pI for GH18 Genes')
plt.xlabel('Molecular Weight (kDa)')
plt.ylabel('Theoretical pI')
plt.show()
```
