This notebook analyzes flow cytometry data to quantify DNA content in cells stained with DAPI.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load DAPI staining data
data = pd.read_csv('dapi_staining_data.csv')

# Analyze and visualize the data
plt.hist(data['DNA_content'], bins=30)
plt.title('Distribution of DNA Content in DAPI Stained Cells')
plt.xlabel('DNA Content (arbitrary units)')
plt.ylabel('Frequency')
plt.show()
```

The histogram visualizes the distribution of DNA content in cells, providing insights into cell cycle phases.
