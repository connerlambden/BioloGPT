### Nutritional Data Analysis for HCT Patients

```python
import pandas as pd

# Load nutritional data
nutritional_data = pd.read_csv('nutritional_data_hct.csv')

# Analyze trends in BMI and MUST scores
display(nutritional_data.describe())
```

### Visualizing Nutritional Trends

```python
import matplotlib.pyplot as plt

# Plotting BMI trends
display(nutritional_data.plot(x='Time', y='BMI', kind='line'))
```

### Correlation Analysis

```python
# Correlation between MUST scores and readmissions
correlation = nutritional_data['MUST'].corr(nutritional_data['Readmissions'])
print(f'Correlation between MUST scores and readmissions: {correlation}')
```
