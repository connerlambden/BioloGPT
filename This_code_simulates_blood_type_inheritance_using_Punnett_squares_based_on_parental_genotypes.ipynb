This notebook analyzes blood type inheritance patterns using genetic data.

```python
# Simulate blood type inheritance
# Define parental genotypes
barbie_genotype = 'ii'  # Type O
ken_genotype = 'IAi'  # Type A (heterozygous)

# Possible offspring genotypes
possible_offspring = [barbie_genotype + ken_genotype]

# Display results
possible_offspring
```

The results show the potential blood types of offspring based on parental genotypes.

```python
# Display possible blood types
blood_types = {'IAi': 'A', 'ii': 'O'}
results = [blood_types[genotype] for genotype in possible_offspring]
results
```
