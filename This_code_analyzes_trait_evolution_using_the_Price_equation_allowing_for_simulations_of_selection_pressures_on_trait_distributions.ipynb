This notebook simulates trait evolution based on the Price equation, incorporating selection pressures and demographic factors.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_generations = 100
population_size = 100
trait_mean = 0
trait_variance = 1

# Simulate trait evolution
traits = np.random.normal(trait_mean, np.sqrt(trait_variance), population_size)
for generation in range(num_generations):
    fitness = np.exp(-traits**2)  # Example fitness function
    traits += np.random.normal(0, 0.1, population_size)  # Trait mutation

# Plot results
plt.hist(traits, bins=20)
plt.title('Trait Distribution After Evolution')
plt.xlabel('Trait Value')
plt.ylabel('Frequency')
plt.show()
```
