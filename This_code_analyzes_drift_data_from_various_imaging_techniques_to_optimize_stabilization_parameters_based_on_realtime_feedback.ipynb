### Drift Analysis for Stabilization Optimization
This notebook analyzes drift data collected from different imaging techniques to optimize stabilization parameters.

```python
import pandas as pd
import matplotlib.pyplot as plt

drift_data = pd.read_csv('drift_data.csv')
plt.figure(figsize=(10,6))
plt.plot(drift_data['time'], drift_data['drift_x'], label='X Drift')
plt.plot(drift_data['time'], drift_data['drift_y'], label='Y Drift')
plt.title('Drift Analysis Over Time')
plt.xlabel('Time (s)')
plt.ylabel('Drift (nm)')
plt.legend()
plt.show()
```
