### Protease Sequence Analysis
This notebook will analyze protease sequences to identify conserved motifs and predict enzyme activity.

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load protease sequence data
protease_data = pd.read_csv('protease_sequences.csv')

# Analyze sequence motifs
motif_analysis = protease_data['sequence'].str.extractall(r'(?P<motif>[A-Z]{3})')

# Visualize motif frequency
sns.countplot(data=motif_analysis, x='motif')
plt.title('Frequency of Motifs in Protease Sequences')
plt.show()
```

This analysis provides insights into the common motifs found in protease sequences, which can be linked to their functional properties.
