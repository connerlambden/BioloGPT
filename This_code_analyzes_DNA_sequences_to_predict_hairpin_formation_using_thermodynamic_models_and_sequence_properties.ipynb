```python
# Function to check for inverted repeats
def has_inverted_repeat(seq):
    for i in range(len(seq) - 1):
        for j in range(i + 2, len(seq) + 1):
            if seq[i:j] == seq[i:j][::-1]:
                return True
    return False

# Check for inverted repeats
inverted_repeat = has_inverted_repeat(str(sequence))
inverted_repeat
```
