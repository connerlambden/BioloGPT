### Data Analysis of CD80 Expression in Heart Macrophages
This notebook will analyze the temporal expression of CD80 in heart macrophages using flow cytometry and immunohistochemistry data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CD80 expression data
cd80_data = pd.read_csv('cd80_expression_data.csv')

# Plotting the data
plt.figure(figsize=(10,6))
plt.plot(cd80_data['Time'], cd80_data['Expression'], marker='o')
plt.title('Temporal Expression of CD80 in Heart Macrophages')
plt.xlabel('Time (hours post-infection)')
plt.ylabel('CD80 Expression Level')
plt.grid(True)
plt.show()
```
