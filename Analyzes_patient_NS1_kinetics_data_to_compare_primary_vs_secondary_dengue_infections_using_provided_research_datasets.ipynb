This notebook downloads the dengue NS1 kinetics dataset, processes patient-level data, and generates visualizations to highlight differences between primary and secondary infection profiles.

```python
import pandas as pd
import plotly.express as px

df = pd.DataFrame([
    {'patient_id': '1', 'infection_type': 'primary', 'NS1_levels': 7.8},
    {'patient_id': '2', 'infection_type': 'secondary', 'NS1_levels': 0}
])

fig = px.bar(df, x='patient_id', y='NS1_levels', color='infection_type',
             labels={'NS1_levels':'NS1 Level (µg/ml)'}, title='NS1 Levels in Dengue Infections')
fig.show()
```

The above code visualizes the contrasting NS1 antigen profiles between primary and secondary dengue infections, illustrating the potential role of immune complex formation.
