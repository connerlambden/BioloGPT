This notebook analyzes neural response data from macaque training studies to identify patterns in symbolic processing capabilities.

```python
import pandas as pd
import numpy as np

# Load the dataset
neural_data = pd.read_csv('macaque_neural_responses.csv')

# Analyze the data
summary_stats = neural_data.describe()
summary_stats
```

The summary statistics provide insights into the neural responses of macaques during training tasks.

```python
# Visualize the data
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(neural_data['response_strength'], bins=30, color='blue', alpha=0.7)
plt.title('Distribution of Neural Response Strengths')
plt.xlabel('Response Strength')
plt.ylabel('Frequency')
plt.show()
```
