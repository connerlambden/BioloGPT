### Docking Analysis of Flunarizine with LASV Proteins
This notebook analyzes the binding affinities of flunarizine to Lassa virus proteins using docking simulations.

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load docking data
# (Assuming docking data is available in a CSV file)
docking_data = pd.read_csv('docking_results.csv')

# Analyze binding affinities
binding_affinities = docking_data['binding_affinity']
mean_affinity = np.mean(binding_affinities)
print(f'Mean Binding Affinity: {mean_affinity}')
```

### Results
The mean binding affinity provides insights into the potential effectiveness of flunarizine against LASV.

```python
# Visualize binding affinities
import matplotlib.pyplot as plt
plt.hist(binding_affinities, bins=20)
plt.title('Distribution of Binding Affinities')
plt.xlabel('Binding Affinity (kcal/mol)')
plt.ylabel('Frequency')
plt.show()
```
