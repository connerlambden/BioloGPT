This notebook will analyze genetic diversity data from SCoT markers, focusing on polymorphism rates and population structure.

```python
import pandas as pd

data = pd.read_csv('scot_marker_data.csv')

# Analyze polymorphism rates
diversity_analysis = data.groupby('marker_type').agg({'polymorphic_bands': 'mean', 'total_bands': 'mean'})
print(diversity_analysis)
```
