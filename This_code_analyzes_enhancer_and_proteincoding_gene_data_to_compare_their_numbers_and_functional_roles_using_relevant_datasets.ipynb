### Data Analysis of Enhancers vs. Protein-Coding Genes
This analysis will utilize genomic datasets to compare the number of enhancers and protein-coding genes in the human genome.

```python
import pandas as pd

# Load datasets
enhancers_data = pd.read_csv('enhancers_data.csv')
protein_coding_data = pd.read_csv('protein_coding_data.csv')

# Count the number of enhancers and protein-coding genes
num_enhancers = len(enhancers_data)
num_protein_coding_genes = len(protein_coding_data)

# Display results
num_enhancers, num_protein_coding_genes
```

### Results
The analysis will provide insights into the relative abundance of enhancers compared to protein-coding genes.

```python
# Display the results
print(f'Number of Enhancers: {num_enhancers}')
print(f'Number of Protein-Coding Genes: {num_protein_coding_genes}')
```
