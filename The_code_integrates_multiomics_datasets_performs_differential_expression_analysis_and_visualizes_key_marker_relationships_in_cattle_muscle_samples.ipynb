This notebook downloads and integrates GWAS, RNA-Seq, and proteomic data relevant to Nellore cattle. It then applies statistical analysis to identify significant markers linking to meat quality traits.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets (paths are placeholders for actual data access links provided by the study)
gwas_data = pd.read_csv('path_to_gwas_data.csv')
transcriptome_data = pd.read_csv('path_to_rnaseq_data.csv')
proteomics_data = pd.read_csv('path_to_proteomics_data.csv')

# Merge datasets on gene symbol
merged_data = pd.merge(transcriptome_data, proteomics_data, on='gene_symbol', how='inner')
merged_data = pd.merge(merged_data, gwas_data, on='gene_symbol', how='inner')

# Differential analysis (example using log2 fold change thresholds)
significant_markers = merged_data[(merged_data['log2FC'] > 1) | (merged_data['log2FC'] < -1)]

# Visualization
plt.figure(figsize=(10,6))
sns.scatterplot(data=significant_markers, x='log2FC', y='p_value', hue='trait', palette='viridis')
plt.title('Significant Differential Markers in Nellore Cattle Muscle')
plt.xlabel('Log2 Fold Change')
plt.ylabel('P-value')
plt.show()
```

The displayed plot visualizes correlations between gene expression changes and p-values across different meat quality traits, aiding in the identification of key regulatory markers.

```python
import plotly.express as px
fig = px.scatter(significant_markers, x='log2FC', y='p_value', color='trait', title='Significant Markers Scatter Plot')
fig.update_layout(title_font_color='#6A0C76')
fig.show()
```

This interactive Plotly graph further allows dynamic exploration of marker significance, which is crucial for hypothesis generation.
