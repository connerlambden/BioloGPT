This notebook analyzes the relationship between urea concentration and bio-oil yield.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Urea Concentration': ['0%', '5%', '10%', '15%', '20%'], 'Bio-oil Yield': [49.33, 53.00, 54.66, 53.33, 52.00]}
df = pd.DataFrame(data)
plt.bar(df['Urea Concentration'], df['Bio-oil Yield'])
plt.title('Bio-oil Yield vs Urea Concentration')
plt.xlabel('Urea Concentration')
plt.ylabel('Bio-oil Yield (%)')
plt.show()
```

The bar chart above illustrates the relationship between urea concentration and bio-oil yield.
