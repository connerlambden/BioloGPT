This notebook will analyze the kinetic energy of divers during jumps using collected data on jump height and time.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for height and time
height = np.array([0, 1, 2, 3, 4, 5])  # height in meters
# Time in seconds
time = np.array([0, 0.5, 1, 1.5, 2, 2.5])

# Calculate kinetic energy at each time point
mass = 70  # average mass of a diver in kg
velocity = np.sqrt(2 * 9.81 * height)  # v = sqrt(2gh)
kinetic_energy = 0.5 * mass * velocity**2

# Plotting the results
plt.plot(time, kinetic_energy)
plt.title('Kinetic Energy of Diver Over Time')
plt.xlabel('Time (s)')
plt.ylabel('Kinetic Energy (J)')
plt.grid()
plt.show()
```

This analysis provides insights into how kinetic energy evolves during a jump, helping to optimize diving techniques.
