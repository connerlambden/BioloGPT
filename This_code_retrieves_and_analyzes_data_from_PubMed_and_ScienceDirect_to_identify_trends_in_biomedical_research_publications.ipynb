### Data Retrieval from PubMed and ScienceDirect

```python
# Import necessary libraries
import requests
import pandas as pd

# Function to retrieve articles from PubMed
def fetch_pubmed_articles(query):
    url = f'https://pubmed.ncbi.nlm.nih.gov/?term={query}'
    response = requests.get(url)
    return response.text

# Function to retrieve articles from ScienceDirect
def fetch_sciencedirect_articles(query):
    url = f'https://www.sciencedirect.com/search?qs={query}'
    response = requests.get(url)
    return response.text
```

### Data Analysis and Visualization

```python
# Analyze and visualize the data
import matplotlib.pyplot as plt

# Sample data for visualization
pubmed_counts = [100, 200, 300]
sciencedirect_counts = [150, 250, 350]

# Create a bar chart
labels = ['2021', '2022', '2023']

x = range(len(labels))
plt.bar(x, pubmed_counts, width=0.4, label='PubMed', color='b', align='center')
plt.bar([p + 0.4 for p in x], sciencedirect_counts, width=0.4, label='ScienceDirect', color='r', align='center')
plt.xlabel('Year')
plt.ylabel('Number of Articles')
plt.title('Comparison of Articles Published in PubMed and ScienceDirect')
plt.xticks([p + 0.2 for p in x], labels)
plt.legend()
plt.show();
```
