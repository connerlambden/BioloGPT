Load and preprocess LC-HRMS dataset for bacterial samples treated with inhibitors.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load dataset
data = pd.read_csv('his_adpr_levels.csv')
# Process and summarize data
summary = data.groupby('treatment')['His_ADPR'].mean()
plt.bar(summary.index, summary.values, color='#6A0C76')
plt.xlabel('Treatment')
plt.ylabel('Mean His-ADPR Level')
plt.title('Effect of Inhibitors on His-ADPR Levels')
plt.show()
```

Analyze significance of His-ADPR reduction and compare with control samples.

```python
import scipy.stats as stats
control = data[data['treatment'] == 'control']['His_ADPR']
treated = data[data['treatment'] == 'inhibitor']['His_ADPR']
stats_result = stats.ttest_ind(control, treated)
print('T-test results:', stats_result)
```
