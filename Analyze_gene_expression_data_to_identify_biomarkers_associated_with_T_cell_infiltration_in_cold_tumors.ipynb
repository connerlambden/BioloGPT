This notebook analyzes gene expression data from cold tumor samples to identify potential biomarkers for T cell infiltration.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_data.csv')
sns.boxplot(x='T_cell_infiltration', y='gene_expression', data=data)
plt.title('T Cell Infiltration vs Gene Expression')
plt.show()
```

The analysis visualizes the relationship between T cell infiltration and gene expression levels, helping to identify potential biomarkers.
