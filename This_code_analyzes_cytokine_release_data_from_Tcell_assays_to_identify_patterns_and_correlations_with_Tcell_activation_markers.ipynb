This notebook will analyze cytokine release data from T-cell assays, focusing on the relationship between cytokine levels and T-cell activation.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cytokine data
data = pd.read_csv('cytokine_data.csv')

# Analyze correlations
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Cytokine Correlation Matrix')
plt.show()
```

The heatmap above illustrates the correlations between different cytokines released during T-cell activation.
