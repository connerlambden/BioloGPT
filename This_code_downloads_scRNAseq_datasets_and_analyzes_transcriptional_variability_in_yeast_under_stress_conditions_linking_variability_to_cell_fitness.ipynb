Retrieve and preprocess the single cell RNA-seq dataset from the provided GEO accession, then partition cells based on stress-induced transcriptional profiles.

```python
import scanpy as sc
adata = sc.read_10x_mtx('path_to_data/')
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.tl.pca(adata, svd_solver='arpack')
sc.pp.neighbors(adata)
sc.tl.umap(adata)
sc.pl.umap(adata, color='stress_marker')
```

Visualize transcriptional heterogeneity with UMAP and identify genetic perturbation effects on transcriptional states.

```python
sc.pl.umap(adata, color=['gene1', 'gene2', 'gene3'], save='_heterogeneity.png')
```

This notebook segments cells by transcriptional response and correlates gene expression variability with experimentally measured fitness.

```python
# Further differential expression analysis
sc.tl.rank_genes_groups(adata, 'stress_state', method='t-test')
sc.pl.rank_genes_groups(adata, n_genes=10, sharey=False)
```
