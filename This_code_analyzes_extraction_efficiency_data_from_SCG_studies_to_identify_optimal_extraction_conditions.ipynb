### Data Analysis of SCG Extraction Efficiency
This notebook will analyze the extraction efficiency of various methods used on spent coffee grounds (SCG) to determine optimal conditions.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for extraction efficiency
extraction_data = {
    'Method': ['60% Methanol', '100% Methanol', '60% Methanol', '100% Methanol'],
    'Total_Phenolic_Content': [16, 2.9, 10.7, 2.6],
    'FRAP': [0.10, 0.036, 0.054, 0.021]
}

df = pd.DataFrame(extraction_data)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.bar(df['Method'], df['Total_Phenolic_Content'], color='blue', alpha=0.6, label='Total Phenolic Content')
plt.bar(df['Method'], df['FRAP'], color='orange', alpha=0.6, label='FRAP')
plt.title('Extraction Efficiency of SCG')
plt.xlabel('Extraction Method')
plt.ylabel('Yield')
plt.legend()
plt.show()
```
