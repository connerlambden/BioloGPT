This notebook will explore genetic data from sympatric populations to identify markers associated with speciation.

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load genetic data
data = pd.read_csv('genetic_data.csv')

# Analyze data for potential speciation markers
markers = data[data['speciation'] == 'sympatric']
markers_summary = markers.describe()
```
