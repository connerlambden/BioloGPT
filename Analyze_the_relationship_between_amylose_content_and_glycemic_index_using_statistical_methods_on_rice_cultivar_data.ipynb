This notebook analyzes the correlation between amylose content and glycemic index in indigenous indica rice cultivars.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'Cultivar': ['Bananthi akki', 'Kaveri', 'Basmathi', 'Neuli'],
        'Glycemic Index': [82.53, 79.30, 78.45, 92.67],
        'Amylose Content': [0.503, 1.296, None, 0.50]}
df = pd.DataFrame(data)
sns.scatterplot(data=df, x='Amylose Content', y='Glycemic Index')
plt.title('Glycemic Index vs. Amylose Content')
plt.show()
```
