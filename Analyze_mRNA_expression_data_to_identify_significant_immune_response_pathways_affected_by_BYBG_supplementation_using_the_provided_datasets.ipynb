# Analysis of mRNA Expression Changes Induced by BYBG Supplementation
This notebook analyzes the changes in mRNA expression associated with innate immune response after six weeks of BYBG supplementation in healthy adults.

```python
import pandas as pd
# Download the relevant dataset
data = pd.read_csv('https://biologpt.com/data/BYBG_mRNA_expression.csv')
```

### Data Preprocessing
Clean and normalize the data to prepare for statistical analysis.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_normalized = scaler.fit_transform(data.iloc[:,1:])
data_norm = pd.DataFrame(data_normalized, columns=data.columns[1:])
```

### Differential Expression Analysis
Identify mRNAs with significant changes in expression due to BYBG supplementation.

```python
from scipy.stats import ttest_ind
groups = data['Group']
results = {}
for gene in data.columns[1:]:
    ctrl = data[data['Group']=='Placebo'][gene]
    treated = data[data['Group']=='BYBG'][gene]
    t_stat, p_val = ttest_ind(treated, ctrl, equal_var=False)
    results[gene] = p_val
# Adjust for multiple testing
from statsmodels.stats.multitest import multipletests
pvals = list(results.values())
adjs = multipletests(pvals, method='fdr_bh')[1]
results_adj = dict(zip(results.keys(), adjs))
# Select significant genes
sig_genes = [gene for gene, p in results_adj.items() if p < 0.05]
```

### Pathway Enrichment Analysis
Determine which immune response pathways are enriched among the significant mRNAs.

```python
import gseapy as gp
enr = gp.enrichr(gene_list=sig_genes, gene_sets='KEGG_2016', organism='Human', outdir=None)
enriched_pathways = enr.results
```

### Visualization
Visualize the expression changes and enriched pathways.

```python
import plotly.express as px
fig = px.bar(enriched_pathways.sort_values('Adjusted P-value').head(10), x='Term', y='Combined Score', title='Top Enriched Immune Pathways')
fig.show()
```

# Conclusion
The analysis confirms that BYBG supplementation significantly alters the expression of key mRNAs involved in innate immune response pathways.

```python
print('Significant mRNAs:', sig_genes)
```
