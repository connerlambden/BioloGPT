This notebook will analyze the effects of various surfactants on alkaline cellulase activity using experimental data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load enzyme activity data
activity_data = pd.read_csv('enzyme_activity_data.csv')

# Analyze and visualize the data
plt.figure(figsize=(10,6))
plt.bar(activity_data['Surfactant'], activity_data['Activity'])
plt.title('Enzyme Activity with Different Surfactants')
plt.xlabel('Surfactant')
plt.ylabel('Activity (U/mL)')
plt.show()
```
