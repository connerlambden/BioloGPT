### Gene Expression Analysis of Plant Response to Ag/rGO Nanocomposites

This notebook analyzes RNA-Seq data from Spinacia oleracea treated with Ag/rGO nanocomposites to identify differentially expressed genes involved in growth regulation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load gene expression data
gene_expr = pd.read_csv('gene_expression.csv')
# Perform differential expression analysis
from scipy.stats import ttest_ind
control = gene_expr[gene_expr['Treatment'] == 'Control']
treated = gene_expr[gene_expr['Treatment'] == 'Ag_rGO']
results = gene_expr[['Gene']].copy()
results['p_value'] = [ttest_ind(control.loc[control['Gene'] == gene, 'Expression'], treated.loc[treated['Gene'] == gene, 'Expression']).pvalue for gene in gene_expr['Gene']]
results['significant'] = results['p_value'] < 0.05
# Save results
genes_sig = results[results['significant']]
genes_sig.to_csv('differentially_expressed_genes.csv', index=False)
```

### Visualization of Differentially Expressed Genes

This section visualizes the genes significantly upregulated or downregulated by Ag/rGO treatment.

```python
# Load significant genes
diff_genes = pd.read_csv('differentially_expressed_genes.csv')
plt.figure(figsize=(10,6))
sns.histplot(diff_genes['p_value'], bins=50, kde=True)
plt.title('P-value Distribution of Differentially Expressed Genes')
plt.xlabel('P-value')
plt.ylabel('Frequency')
plt.savefig('pvalue_distribution.png')
plt.show()
```

### Conclusion

The analysis identifies key genes modulated by Ag/rGO nanocomposites, providing insights into their role in enhancing plant growth.
