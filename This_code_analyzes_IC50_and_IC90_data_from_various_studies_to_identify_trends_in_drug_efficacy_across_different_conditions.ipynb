### Data Analysis of IC50 and IC90 Values
This notebook will analyze IC50 and IC90 values from selected studies to identify trends and correlations.

```python
import pandas as pd

# Load IC50 and IC90 data from studies
data = pd.read_csv('ic_data.csv')

# Analyze the data
summary = data.describe()
summary
```

### Summary of IC50 and IC90 Data
The summary provides insights into the average potency of drugs across different studies.

```python
# Visualize the data
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(data['IC50'], data['IC90'])
plt.title('IC50 vs IC90 Analysis')
plt.xlabel('IC50 (µg/ml)')
plt.ylabel('IC90 (µg/ml)')
plt.show()
```
