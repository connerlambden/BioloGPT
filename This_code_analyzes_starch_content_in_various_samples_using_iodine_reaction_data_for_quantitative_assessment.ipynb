### Starch Content Analysis
This notebook will analyze the starch content in various samples based on the iodine reaction data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for starch testing
samples = {'Sample': ['Sample A', 'Sample B', 'Sample C'], 'Color Change': ['Blue-Black', 'No Change', 'Blue-Black']}
df = pd.DataFrame(samples)

# Count positive tests
positive_tests = df[df['Color Change'] == 'Blue-Black'].count()['Sample']

# Plotting results
plt.bar(df['Sample'], df['Color Change'].apply(lambda x: 1 if x == 'Blue-Black' else 0))
plt.title('Starch Test Results')
plt.xlabel('Samples')
plt.ylabel('Positive Test (1 = Yes, 0 = No)')
plt.show()
```

### Results Interpretation
The bar chart visualizes the positive test results for starch in the samples.

```python
# Displaying the number of positive tests
print(f'Number of positive tests: {positive_tests}')
```
