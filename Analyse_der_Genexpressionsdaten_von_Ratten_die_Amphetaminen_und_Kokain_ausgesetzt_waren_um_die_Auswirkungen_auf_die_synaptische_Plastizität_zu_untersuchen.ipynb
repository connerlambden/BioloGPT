Daten zu Genexpressionsänderungen nach Drogenexposition analysieren.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Daten laden
data = pd.read_csv('gene_expression_data.csv')

# Daten visualisieren
sns.boxplot(x='treatment', y='gene_expression', data=data)
plt.title('Genexpressionsänderungen nach Drogenexposition')
plt.show()
```

Die Analyse zeigt, wie Amphetamine und Kokain die Genexpression in neuronalen Zellen beeinflussen.
