This notebook will analyze V4 neural response data in relation to foraging behavior metrics.

```python
import pandas as pd
import numpy as np

# Load V4 neural response data
v4_data = pd.read_csv('v4_neural_responses.csv')

# Load foraging behavior data
foraging_data = pd.read_csv('foraging_behavior.csv')

# Merge datasets on monkey ID
merged_data = pd.merge(v4_data, foraging_data, on='monkey_id')

# Analyze correlation between contrast invariance and foraging success
correlation = merged_data['contrast_invariance'].corr(merged_data['foraging_success'])
print('Correlation between contrast invariance and foraging success:', correlation)
```
