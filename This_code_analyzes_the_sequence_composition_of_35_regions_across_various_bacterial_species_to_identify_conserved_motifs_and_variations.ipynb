### Analyzing -35 Sequence Variations
This notebook will analyze the -35 sequences from a dataset of bacterial promoters to identify common motifs and variations.

```python
import pandas as pd

# Load dataset containing -35 sequences
sequences = pd.read_csv('bacterial_promoter_sequences.csv')

# Analyze the length and composition of -35 sequences
sequences['length'] = sequences['-35_sequence'].apply(len)
common_motifs = sequences['-35_sequence'].value_counts().head(10)

common_motifs
```

### Results
The analysis reveals the most common -35 sequences and their frequencies across the dataset.

```python
# Display the common motifs
common_motifs.plot(kind='bar', title='Top 10 Common -35 Sequences')
```
