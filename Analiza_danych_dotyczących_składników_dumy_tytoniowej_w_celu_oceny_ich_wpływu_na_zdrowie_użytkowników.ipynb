Analiza danych dotyczących składników tytoniowych i ich wpływu na zdrowie.

```python
import pandas as pd
# Wczytanie danych o składnikach
skladniki = pd.read_csv('skladniki_tytoniu.csv')
# Analiza danych
skladniki.describe()
```
