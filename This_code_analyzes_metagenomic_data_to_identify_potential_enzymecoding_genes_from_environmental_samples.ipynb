### Step 1: Import necessary libraries and load metagenomic data.

```python
import pandas as pd
import numpy as np

# Load metagenomic data
metagenomic_data = pd.read_csv('metagenomic_data.csv')
```

### Step 2: Filter data for enzyme-coding genes.

```python
# Filter for enzyme-coding genes
enzyme_genes = metagenomic_data[metagenomic_data['function'] == 'enzyme']
```

### Step 3: Analyze the abundance of enzyme genes.

```python
# Analyze abundance
abundance = enzyme_genes.groupby('gene_id').size().reset_index(name='counts')

# Display results
abundance.sort_values(by='counts', ascending=False).head(10)
```
