This notebook will analyze datasets from adaptive evolution experiments to determine the effects of selective pressures.

```python
import pandas as pd
import numpy as np

# Load datasets
pressure_data = pd.read_csv('pressure_data.csv')

# Analyze fitness changes under different pressures
fitness_analysis = pressure_data.groupby(['pressure_type']).agg({'fitness': 'mean'})

# Output results
fitness_analysis.to_csv('fitness_analysis.csv')
```

The analysis provides insights into how different selective pressures affect fitness, helping to identify orthogonal interactions.

```python
# Visualize results
import matplotlib.pyplot as plt

plt.bar(fitness_analysis.index, fitness_analysis['fitness'])
plt.title('Fitness under Different Selective Pressures')
plt.xlabel('Selective Pressure Type')
plt.ylabel('Mean Fitness')
plt.show()
```
