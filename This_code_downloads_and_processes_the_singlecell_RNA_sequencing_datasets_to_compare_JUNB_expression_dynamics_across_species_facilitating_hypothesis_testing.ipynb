Collect and preprocess single-cell RNA-seq datasets from mouse, human neocortex, and human organoids using Seurat and custom R scripts. Visualize JUNB expression across clusters.

```python
import scanpy as sc
adata_mouse = sc.read_h5ad('mouse_neocortex.h5ad')
adata_human = sc.read_h5ad('human_neocortex.h5ad')
adata_organoids = sc.read_h5ad('human_organoids.h5ad')

# Normalize and integrate datasets
sc.pp.normalize_total(adata_mouse)
sc.pp.log1p(adata_mouse)
# ... similar processing for human datasets

# Visualize JUNB expression
sc.pl.umap(adata_human, color='JUNB', save='_human_JUNB_expression.png')
sc.pl.umap(adata_mouse, color='JUNB', save='_mouse_JUNB_expression.png')
```

The notebook performs integration of datasets, normalization, and UMAP visualization to directly compare JUNB levels across species.

```python
# Further statistical analysis to compare JUNB expression across groups
import pandas as pd
import seaborn as sns

df = pd.concat([adata_mouse.to_df()[['JUNB']], adata_human.to_df()[['JUNB']]], axis=0)
sns.boxplot(data=df)
```
