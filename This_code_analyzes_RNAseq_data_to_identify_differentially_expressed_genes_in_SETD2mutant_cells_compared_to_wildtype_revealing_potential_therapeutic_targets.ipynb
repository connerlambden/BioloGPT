This notebook analyzes RNA-seq data from SETD2-mutant and wild-type cells.

```python
import pandas as pd
import numpy as np

# Load RNA-seq data
data = pd.read_csv('RNA_seq_data.csv')

# Identify differentially expressed genes
diff_expr = data[data['p_value'] < 0.05]
```

The analysis identifies genes that are significantly altered in expression, which may serve as potential targets for therapy.

```python
# Output the results
diff_expr.to_csv('differentially_expressed_genes.csv', index=False)
```
