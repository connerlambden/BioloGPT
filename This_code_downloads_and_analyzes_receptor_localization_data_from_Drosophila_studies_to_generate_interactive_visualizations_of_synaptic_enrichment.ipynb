This notebook downloads receptor enrichment data from public datasets and processes it to generate visualizations that illustrate the differential localization of Dop1R1 and Dop2R.

```python
import pandas as pd
import plotly.express as px
# Load dataset (replace URL with actual data source URL)
df = pd.read_csv('https://doi.org/10.12751/g-node.i3xmim')
fig = px.bar(df, x='Receptor', y='Enrichment', color='Receptor', title='Presynaptic Enrichment of Dopamine Receptors')
fig.show()
```

The code uses the Plotly library to generate an interactive bar graph that displays the relative presynaptic enrichment values of the two dopamine receptors.

```python
# Additional analysis can include statistical tests to compare receptor distributions
import scipy.stats as stats
stat, p = stats.ttest_ind(df[df['Receptor']=='Dop1R1']['Enrichment'], df[df['Receptor']=='Dop2R']['Enrichment'])
print('T-test statistic:', stat, 'p-value:', p)
```

This additional block performs a t-test to determine if the observed differences in enrichment are statistically significant.
