### Genetic Diversity Analysis of Arabidopsis Arenosa
This notebook analyzes the genetic diversity of diploid and tetraploid populations of Arabidopsis arenosa using sequencing data.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Ploidy': ['Diploid', 'Tetraploid'], 'Nucleotide Diversity': [0.021, 0.025]}
df = pd.DataFrame(data)

plt.bar(df['Ploidy'], df['Nucleotide Diversity'], color=['blue', 'green'])
plt.title('Nucleotide Diversity in Arabidopsis')
plt.ylabel('Nucleotide Diversity (π)')
plt.show()
```

### Results
The bar chart above illustrates the differences in nucleotide diversity between diploid and tetraploid populations.
