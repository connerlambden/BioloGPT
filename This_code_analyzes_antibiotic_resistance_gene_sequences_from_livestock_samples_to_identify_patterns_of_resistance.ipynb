This notebook will analyze the prevalence of antibiotic resistance genes in livestock samples.

```python
import pandas as pd
# Load dataset
data = pd.read_csv('antibiotic_resistance_genes.csv')
# Analyze data
resistance_patterns = data['gene'].value_counts()
```

The analysis reveals the most common antibiotic resistance genes present in the samples.

```python
resistance_patterns.plot(kind='bar')
```
