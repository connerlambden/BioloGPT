### Data Preparation
Load the sequencing data and prepare it for analysis.

```python
import pandas as pd

# Load sequencing data
sequencing_data = pd.read_csv('sequencing_data.csv')

# Display the first few rows
sequencing_data.head()
```

### Analysis of Selection Efficiency
Calculate the enrichment of target sequences after each selection cycle.

```python
# Calculate enrichment
enrichment = sequencing_data['target_count'] / sequencing_data['total_count']

# Add enrichment to the dataframe
sequencing_data['enrichment'] = enrichment

# Display the updated dataframe
sequencing_data.head()
```

### Visualization
Visualize the enrichment across selection cycles.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(sequencing_data['cycle'], sequencing_data['enrichment'], marker='o')
plt.title('Enrichment of Target Sequences Across Selection Cycles')
plt.xlabel('Selection Cycle')
plt.ylabel('Enrichment Ratio')
plt.grid()
plt.show()
```
