This section describes the steps to load TCGA methylation datasets, preprocess the data, compute the percentage of 2-allele demethylation, and perform statistical associations with cancer outcomes.

```python
import pandas as pd
import numpy as np
from statsmodels.formula.api import ols

# Load methylation data
methylation_data = pd.read_csv('path/to/tcga_methylation.csv')

# Compute X-Ra: percent of 2-allele demethylation
methylation_data['X_Ra'] = methylation_data.filter(like='XCI_').apply(lambda row: np.mean(row==0), axis=1)

# Correlate with cancer outcomes
model = ols('cancer_outcome ~ X_Ra', data=methylation_data).fit()
print(model.summary())
```

This code snippet demonstrates the statistical association between the computed X-Ra values and cancer clinical outcomes, providing a replicable analysis workflow.
