This notebook processes the simulation trajectories from GROMACS output to map critical contact residues between the thioredoxin fold of SH3BGRL and actin.

```python
import MDAnalysis as mda
import ProLIF
# Load trajectory and topology files
u = mda.Universe('topology.pdb', 'trajectory.xtc')
contacts = ProLIF.algo.distance_contacts(u, selection='protein', cutoff=0.5)
print(contacts)
```

The code above calculates residue-residue contacts and outputs a contact map, assisting in pinpointing interaction hotspots.

```python
# Visualization code using Plotly
import plotly.express as px
fig = px.imshow(contacts, color_continuous_scale='Viridis', title='SH3BGRL-Actin Contact Map')
fig.show()
```
