This notebook will analyze ingredient lists from baby wipe products to identify potential allergens and irritants.

```python
import pandas as pd

# Load ingredient data
ingredients = pd.read_csv('baby_wipes_ingredients.csv')

# Analyze common allergens
allergens = ['methylisothiazolinone', 'fragrance', 'alcohol']
common_allergens = ingredients[ingredients['ingredient'].isin(allergens)]
common_allergens.groupby('brand').size()
```

The analysis will provide insights into which brands contain the most common allergens.
