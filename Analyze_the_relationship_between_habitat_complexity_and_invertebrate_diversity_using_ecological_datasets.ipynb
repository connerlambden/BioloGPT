This notebook will explore the correlation between habitat complexity metrics and invertebrate diversity indices using available ecological datasets.

```python
import pandas as pd

# Load datasets
complexity_data = pd.read_csv('habitat_complexity.csv')
diversity_data = pd.read_csv('invertebrate_diversity.csv')

# Merge datasets on common keys
merged_data = pd.merge(complexity_data, diversity_data, on='site_id')

# Analyze correlation
correlation = merged_data.corr()['complexity_index']['diversity_index']
print(f'Correlation between habitat complexity and invertebrate diversity: {correlation}')
```

The analysis reveals the strength of the relationship between habitat complexity and invertebrate diversity, providing insights for conservation strategies.

```python
# Visualize the relationship
import matplotlib.pyplot as plt
plt.scatter(merged_data['complexity_index'], merged_data['diversity_index'])
plt.title('Habitat Complexity vs Invertebrate Diversity')
plt.xlabel('Habitat Complexity Index')
plt.ylabel('Invertebrate Diversity Index')
plt.show()
```
