Step-by-step instructions to load and normalize raw qPCR data from infected bird and vector samples.

```python
import pandas as pd
import numpy as np
# Code to load qPCR dataset
df = pd.read_csv('qPCR_data.csv')
# Normalize and process data
df['normalized'] = df['Ct'] / np.mean(df['Ct'])
print(df.head())
```
