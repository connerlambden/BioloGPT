This notebook will analyze gene expression data from bronchiectasis patients to identify correlations between pKLC102 presence and exacerbation severity.

```python
import pandas as pd
# Load gene expression data
data = pd.read_csv('gene_expression_data.csv')
# Analyze correlations with exacerbation severity
correlation_results = data.corr()
```
