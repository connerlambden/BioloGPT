Análisis de datos de proyectos ambientales en escuelas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos de proyectos ambientales
data = pd.read_csv('proyectos_ambientales.csv')

# Visualizar resultados
plt.figure(figsize=(10,6))
plt.bar(data['Proyecto'], data['Impacto'])
plt.title('Impacto de Proyectos Ambientales en Escuelas')
plt.xlabel('Proyectos')
plt.ylabel('Impacto')
plt.show()
```

Este análisis permite visualizar el impacto de diferentes proyectos ambientales en las escuelas.
