This notebook will guide you through the analysis of ecotoxicological data, focusing on species sensitivity and modes of action.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('ecotoxicology_data.csv')
sns.boxplot(x='chemical', y='LC50', data=data)
plt.title('Species Sensitivity to Chemicals')
plt.show()
```

The above code visualizes the sensitivity of different species to various chemicals, helping to identify patterns in toxicity.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = data[['chemical', 'concentration']]
y = data['response']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = RandomForestClassifier()
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print(f'Model Accuracy: {accuracy}')
```

This section trains a machine learning model to predict responses based on chemical exposure, providing insights into modes of action.
