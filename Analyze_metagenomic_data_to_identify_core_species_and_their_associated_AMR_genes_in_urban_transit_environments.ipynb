```python
import pandas as pd

# Load metagenomic data
data = pd.read_csv('urban_transit_metagenomics.csv')

# Analyze core species and AMR genes
core_species = data[data['is_core'] == True]

# Output results
core_species.to_csv('core_species_amr.csv', index=False)
```
