This notebook will analyze GPS tracking data of elk to determine how wolf presence influences their movement patterns.

```python
import pandas as pd

# Load elk movement data
data = pd.read_csv('elk_movement_data.csv')

# Analyze movement patterns based on wolf presence
results = data.groupby('wolf_presence').agg({'movement_distance': 'mean'})
results
```

The results will show the average movement distance of elk in areas with and without wolf presence.

```python
results.to_csv('elk_movement_analysis.csv')
```
