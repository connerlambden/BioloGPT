Load and preprocess single-cell RNA-seq data from patient-derived TIL samples, with a focus on markers like granzyme B, perforin, and exhaustion genes.

```python
import scanpy as sc
adata = sc.read_h5ad('TIL_dataset.h5ad')
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.tl.pca(adata, svd_solver='arpack')
sc.pl.pca(adata, color=['GranzymeB', 'Perforin'], save='PCA_plot.png')
```

The code above normalizes single-cell data, performs PCA, and visualizes key cytotoxic markers to elucidate differences between low- and high-expanding TIL subsets.

```python
sc.tl.umap(adata)
sc.pl.umap(adata, color=['TIL_subset', 'Exhaustion_marker'], save='UMAP_TIL.png')
```

Subsequent UMAP analysis enables cluster identification that may correlate with the expansion profiles observed in the SIN-treated cultures.
