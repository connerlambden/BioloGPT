### Analysis of Grease Barrier Properties of Packaging Paper
This notebook analyzes the effectiveness of different coatings on the grease barrier properties of packaging paper using experimental data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Coating Type': ['Base Paper', 'PVA Coated', 'Kaolin Coated', 'NFC Coated'],
    'WVTR (g/m²•day)': [533, 1.3, 15, 0],
    'Grease Resistance Kit Value': [0, 12, 12, 12]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Bar plot for WVTR
ax[0].bar(df['Coating Type'], df['WVTR (g/m²•day)'], color='blue')
ax[0].set_title('WVTR of Coated Papers')
ax[0].set_ylabel('WVTR (g/m²•day)')

# Bar plot for Grease Resistance
ax[1].bar(df['Coating Type'], df['Grease Resistance Kit Value'], color='green')
ax[1].set_title('Grease Resistance of Coated Papers')
ax[1].set_ylabel('Grease Resistance Kit Value')

plt.tight_layout()
plt.show()
```

This analysis highlights the significant improvements in grease barrier properties achieved through various coatings.
