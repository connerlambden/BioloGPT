This notebook will analyze blaTEM sequences from various isolates to identify mutations that correlate with resistance levels.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load blaTEM sequence data
blaTEM_data = pd.read_csv('blaTEM_sequences.csv')

# Analyze mutations and resistance levels
mutation_analysis = blaTEM_data.groupby('mutation').mean()['resistance_level']

# Visualize the results
plt.figure(figsize=(10,6))
sns.barplot(x=mutation_analysis.index, y=mutation_analysis.values)
plt.title('Mutation Impact on blaTEM Resistance Levels')
plt.xlabel('Mutations')
plt.ylabel('Average Resistance Level')
plt.show()
```

This analysis provides insights into which mutations in the blaTEM gene are most strongly associated with increased resistance, guiding future research.
