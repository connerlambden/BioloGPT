### Genetic Diversity Analysis of Plant Species
This notebook analyzes the genetic diversity of aspen and willow populations in relation to herbivore browsing patterns influenced by wolves.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load genetic diversity data
aspen_data = pd.read_csv('aspen_genetic_diversity.csv')
willow_data = pd.read_csv('willow_genetic_diversity.csv')

# Analyze and visualize the data
plt.figure(figsize=(10, 5))
plt.plot(aspen_data['Year'], aspen_data['Genetic_Diversity'], label='Aspen Genetic Diversity')
plt.plot(willow_data['Year'], willow_data['Genetic_Diversity'], label='Willow Genetic Diversity')
plt.title('Genetic Diversity Over Time')
plt.xlabel('Year')
plt.ylabel('Genetic Diversity Index')
plt.legend()
plt.show()
```

### Results
The analysis reveals trends in genetic diversity for both aspen and willow populations over time, correlating with changes in herbivore browsing pressure.
