This notebook downloads vesicle metrics data from the paper and visualizes the differences in size, count, and density between dopaminergic and glutamatergic synaptosomes.

```python
import plotly.graph_objs as go

# Data from research paper
metrics = ['Mean Size (nm)', 'Average Vesicle Count', 'Vesicle Density (vesicles/µm³)']
da_values = [575, 30, 723]
glu_values = [823, 191, 1956]

trace1 = go.Bar(x=metrics, y=da_values, name='DA Synaptosomes', marker=dict(color='#6A0C76'))
trace2 = go.Bar(x=metrics, y=glu_values, name='GLU Synaptosomes', marker=dict(color='#FFA500'))

data = [trace1, trace2]

layout = go.Layout(title='Vesicle Metrics in DA vs GLU Synaptosomes', barmode='group')

fig = go.Figure(data=data, layout=layout)
fig.show()
```

The graph clearly contrasts the metrics, highlighting the significant organizational differences between the two synaptosome types.
