### Phylogenetic Analysis of EsHV1
This notebook performs a phylogenetic analysis of EsHV1 based on DNA polymerase and terminase gene sequences to ascertain its genetic relationships within the Alloherpesviridae family.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from Bio import Phylo
from io import StringIO

# Load phylogenetic tree data (Newick format)
tree_data = '''(EsHV1_swe, (Other_Viruses));'''  # Replace with actual data

# Parse the tree
tree = Phylo.read(StringIO(tree_data), 'newick')

# Plot the tree
Phylo.draw(tree, do_show=False)
plt.title('Phylogenetic Placement of EsHV1')
plt.savefig('phylogenetic_tree.png')
plt.show()
```
