This notebook will analyze RNA-seq data to correlate mRNA expression levels with SFPQ puncta sizes.

```python
import pandas as pd
# Load RNA-seq data and SFPQ puncta size data
rna_seq_data = pd.read_csv('rna_seq_data.csv')
puncta_data = pd.read_csv('puncta_size_data.csv')
# Merge datasets and analyze correlations.
```

The analysis will focus on identifying mRNAs that show significant expression changes correlated with SFPQ puncta size.

```python
# Perform correlation analysis
correlation_results = rna_seq_data.corrwith(puncta_data['puncta_size'])
# Output results
correlation_results.to_csv('correlation_results.csv')
```
