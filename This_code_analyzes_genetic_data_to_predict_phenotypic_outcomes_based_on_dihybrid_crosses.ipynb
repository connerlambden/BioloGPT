This notebook will analyze genetic data from dihybrid crosses to predict phenotypic outcomes.

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Define the genotypes and their corresponding phenotypes
genotypes = {'GGEE': 'Curly, Brown', 'ggee': 'Straight, Blue'}

# Function to predict offspring phenotype

def predict_offspring(parent1, parent2):
    return genotypes[parent1], genotypes[parent2]

# Example prediction
predict_offspring('GGEE', 'ggee')
```
