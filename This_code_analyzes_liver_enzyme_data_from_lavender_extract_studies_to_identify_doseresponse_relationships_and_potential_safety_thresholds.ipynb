### Data Analysis of Lavender Extract Effects on Liver Enzymes

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Dose (mg/kg)': [21.1, 211, 2110], 'ALT (U/L)': [30, 35, 60], 'AST (U/L)': [25, 28, 50]}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 5))
plt.plot(df['Dose (mg/kg)'], df['ALT (U/L)'], marker='o', label='ALT')
plt.plot(df['Dose (mg/kg)'], df['AST (U/L)'], marker='o', label='AST')
plt.xscale('log')
plt.title('Liver Enzyme Levels vs. Lavender Extract Dose')
plt.xlabel('Dose (mg/kg)')
plt.ylabel('Enzyme Level (U/L)')
plt.legend()
plt.grid()
plt.show()
```
