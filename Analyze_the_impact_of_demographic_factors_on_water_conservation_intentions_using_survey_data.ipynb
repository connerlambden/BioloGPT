This notebook analyzes survey data to understand how demographic factors influence water conservation intentions.

```python
import pandas as pd

# Load survey data
data = pd.read_csv('water_conservation_survey.csv')

# Analyze the impact of demographics on intentions
results = data.groupby('demographic_factor')['intention'].mean()
```

The results show the average intention to conserve water across different demographic groups.

```python
results.to_csv('demographic_impact_analysis.csv')
```
