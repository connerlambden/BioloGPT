This notebook analyzes gene expression data from muscle contraction studies to identify differentially expressed genes and their roles in muscle physiology.

```python
import pandas as pd
import numpy as np

# Load gene expression data
gene_data = pd.read_csv('gene_expression_data.csv')

# Identify differentially expressed genes
differential_genes = gene_data[gene_data['p_value'] < 0.05]

# Output results
differential_genes.to_csv('differential_genes.csv', index=False)
```
