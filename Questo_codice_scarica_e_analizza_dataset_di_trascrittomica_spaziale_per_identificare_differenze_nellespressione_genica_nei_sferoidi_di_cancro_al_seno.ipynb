### Analisi dei Dataset di Trascrittomica Spaziale
Iniziamo scaricando i dataset pertinenti e analizzando l'espressione genica.

```python
import scanpy as sc

# Carica i dati da un file di esempio
adata = sc.read('path_to_dataset.h5ad')

# Esegui analisi di qualità
sc.pp.filter_genes(adata, min_cells=3)
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata) 

# Visualizza i risultati
sc.pl.highest_expr_genes(adata, n_top=20)
```
