### Data Analysis of LBP and suPAR Levels
This notebook will analyze the levels of LBP and suPAR in patient samples to assess their diagnostic value.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for LBP and suPAR levels
data = {'Group': ['STB Patients', 'Healthy Controls'], 'LBP': [102.73, 81.3], 'suPAR': [10.0, 6.2]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Group'], df['LBP'], color='blue', label='LBP Levels (ng/mL)')
plt.bar(df['Group'], df['suPAR'], color='orange', label='suPAR Levels (ng/mL)', alpha=0.7)
plt.title('Comparison of LBP and suPAR Levels')
plt.ylabel('Levels (ng/mL)')
plt.legend()
plt.show()
```

This analysis provides insights into the comparative levels of LBP and suPAR in different patient groups.
