Análisis de datos de expresión génica durante la tianesis.

```python
import pandas as pd
# Cargar datos de expresión génica
exp_data = pd.read_csv('data/tianesis_expression.csv')
exp_data.head()
```

Análisis de la expresión génica en diferentes etapas de la tianesis.

```python
# Análisis de expresión
exp_analysis = exp_data.groupby('stage').mean()
exp_analysis.plot(kind='bar')
```

Visualización de los resultados.

```python
import matplotlib.pyplot as plt
plt.show()
```
