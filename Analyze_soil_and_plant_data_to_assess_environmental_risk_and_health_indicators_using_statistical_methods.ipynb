### Soil and Plant Data Analysis
This notebook will analyze the soil and plant data collected from the agrosilvopastoral system.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Parameter': ['Cu', 'Mo', 'Zn', 'Fe', 'K'],
    'Concentration': [220, 8.2, 350, 660.5, 21.1]
}
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Parameter'], df['Concentration'])
plt.title('Soil and Plant Metal Concentrations')
plt.xlabel('Elements')
plt.ylabel('Concentration (mg/kg or g/kg)')
plt.show()
```
