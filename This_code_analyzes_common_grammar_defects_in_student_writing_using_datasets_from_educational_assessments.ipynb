### Grammar Defects Analysis
This notebook analyzes the frequency and types of grammar defects found in student writing.

```python
import pandas as pd

# Load data
errors_data = pd.read_csv('grammar_errors.csv')

# Analyze data
error_summary = errors_data.groupby('error_type').size()
error_summary
```

### Summary of Errors
The analysis provides insights into the most common types of grammar defects.

```python
error_summary.plot(kind='bar', title='Common Grammar Defects')
```
