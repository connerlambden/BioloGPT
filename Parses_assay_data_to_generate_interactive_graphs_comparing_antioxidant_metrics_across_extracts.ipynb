This notebook downloads the antioxidant assay dataset and visualizes key parameters, such as phenolic content and DPPH IC50 values.

```python
import plotly.express as px
import pandas as pd

data = pd.DataFrame([
    {'plant': 'Coffea arabica', 'phenolic': 1378.19, 'DPPH_IC50': 0.17},
    {'plant': 'Curcuma longa', 'phenolic': 514.13, 'DPPH_IC50': 0.42},
    {'plant': 'Bacopa monnieri', 'phenolic': 258.02, 'DPPH_IC50': 0.38},
    {'plant': 'Camellia sinensis', 'phenolic': 387.43, 'DPPH_IC50': 0.55},
    {'plant': 'Tagetes erecta', 'phenolic': None, 'DPPH_IC50': 12.40},
    {'plant': 'Terminalia chebula', 'phenolic': None, 'DPPH_IC50': 0.98}
])

fig = px.bar(data, x='plant', y='phenolic', title='Total Phenolic Content Comparison', labels={'phenolic':'mg GAE/g'})
fig.show()
```

The code above visualizes the total phenolic content across the examined extracts. Similar plots can be created for other antioxidant parameters.

```python
fig2 = px.bar(data, x='plant', y='DPPH_IC50', title='DPPH IC50 Values', labels={'DPPH_IC50':'mg/mL'})
fig2.show()
```
