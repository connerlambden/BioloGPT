Step-by-step guide for analyzing fMRI activation in hippocampal ROI during episodic versus schematic mismatches using Nilearn.

```python
import nilearn
from nilearn import image, plotting
# Load preprocessed fMRI images for episodic and schematic conditions
episodic_img = image.load_img('path_to_ep_data.nii')
schematic_img = image.load_img('path_to_schema_data.nii')
```

Compute the contrast image by subtracting schematic condition from episodic condition and visualize the result.

```python
contrast_img = image.math_img('img1 - img2', img1=episodic_img, img2=schematic_img)
plotting.plot_stat_map(contrast_img, title='Hippocampal Mismatch Contrast')
plotting.show()
```

This analysis allows researchers to visualize which areas of the hippocampus are preferentially activated during episodic memory-based prediction errors.
