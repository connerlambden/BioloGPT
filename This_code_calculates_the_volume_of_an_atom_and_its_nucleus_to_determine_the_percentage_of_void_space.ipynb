### Volume Calculation of Atom and Nucleus
This section calculates the volumes of an atom and its nucleus to determine the percentage of void.

```python
import math

# Constants
radius_atom = 5.3e-11  # meters (hydrogen atom)
radius_nucleus = 1.2e-15  # meters (nucleus)

# Volume calculations
V_atom = (4/3) * math.pi * (radius_atom ** 3)
V_nucleus = (4/3) * math.pi * (radius_nucleus ** 3)

# Percentage of void
percentage_void = (1 - (V_nucleus / V_atom)) * 100
percentage_void
```

The calculated percentage of void in the atom will be displayed.
