This notebook will analyze the sentiment of selected poems by Giosuè Carducci.

```python
import pandas as pd
from textblob import TextBlob

# Load Carducci's poems
poems = pd.read_csv('carducci_poems.csv')

# Function to calculate sentiment
poems['sentiment'] = poems['text'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Display results
poems[['title', 'sentiment']]
```

The sentiment analysis results will provide insights into the emotional landscape of Carducci's poetry.

```python
poems.groupby('title')['sentiment'].mean().plot(kind='bar', title='Average Sentiment of Carducci's Poems')
```
