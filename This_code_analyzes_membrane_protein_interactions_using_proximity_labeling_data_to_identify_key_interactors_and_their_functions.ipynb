### Membrane Protein Interaction Analysis
This notebook analyzes membrane protein interactions based on proximity labeling data.

```python
import pandas as pd

# Load proximity labeling data
data = pd.read_csv('proximity_labeling_data.csv')

# Analyze interactions
interaction_summary = data.groupby('protein').agg({'interactor': 'count'})
interaction_summary.sort_values(by='interactor', ascending=False).head(10) # Top 10 interactors
```
