This notebook loads the participant data, processes FIM/WeeFIM scores and satisfaction categories, and generates plots for further analysis.

```python
import pandas as pd
import plotly.express as px
# Load dataset
data = [{'Participant_ID': 1, 'Age': 37.5, 'Height': 139, 'NDIS_Access': 'Yes', 'Satisfaction': 'Satisfied', 'FIM_Score': 115.5},
        {'Participant_ID': 2, 'Age': 8.5, 'Height': 105, 'NDIS_Access': 'Yes', 'Satisfaction': 'Extremely satisfied', 'WeeFIM_Score': 95.5}]
df = pd.DataFrame(data)
# Create scatter plot
fig = px.scatter(df, x='Participant_ID', y='FIM_Score' if 'FIM_Score' in df.columns else 'WeeFIM_Score', color='Satisfaction', title='Functional Scores by Participant')
fig.show()
```
