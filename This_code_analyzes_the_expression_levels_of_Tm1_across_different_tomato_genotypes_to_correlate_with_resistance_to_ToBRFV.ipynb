This notebook will analyze Tm-1 expression data from various tomato genotypes.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('tm1_expression_data.csv')
plt.figure(figsize=(10,6))
plt.bar(data['Genotype'], data['Expression_Level'])
plt.title('Tm-1 Expression Levels in Tomato Genotypes')
plt.xlabel('Genotype')
plt.ylabel('Expression Level')
plt.show()
```
