Analisis data genetik dari spesies Artocarpus untuk mengeksplorasi variasi dalam karakteristik buah.

```python
import pandas as pd

# Mengimpor dataset genetik dari Artocarpus
dataset = pd.read_csv('artocarpus_genetic_data.csv')

# Menampilkan ringkasan dataset
dataset.describe()
```

Analisis ini akan membantu dalam memahami variasi genetik dan potensi aplikasi dalam pertanian.

```python
# Menghitung variasi genetik
variations = dataset['genetic_marker'].value_counts()
variations.plot(kind='bar')
```
