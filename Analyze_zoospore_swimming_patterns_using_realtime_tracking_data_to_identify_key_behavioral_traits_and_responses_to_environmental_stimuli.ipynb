This notebook will analyze the swimming trajectories of zoospores, focusing on their response to chemical gradients.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load zoospore swimming data
data = pd.read_csv('zoospore_swimming_data.csv')

# Analyze swimming speed and direction changes
plt.figure(figsize=(10, 6))
plt.plot(data['time'], data['speed'], label='Swimming Speed')
plt.xlabel('Time (s)')
plt.ylabel('Speed (µm/s)')
plt.title('Zoospore Swimming Speed Over Time')
plt.legend()
plt.show()
```

This analysis will help visualize how zoospore swimming behavior changes in response to environmental factors.

```python
# Further analysis of direction changes
plt.figure(figsize=(10, 6))
plt.hist(data['direction_change'], bins=30, alpha=0.7, color='blue')
plt.xlabel('Direction Change (degrees)')
plt.ylabel('Frequency')
plt.title('Distribution of Direction Changes in Zoospore Swimming')
plt.show()
```
