### Toxicity Analysis of Aluminum Nanoparticles
This notebook will analyze the toxicity data of aluminum nanoparticles on aquatic organisms and assess correlations with environmental factors.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Nanoparticle': ['Aluminum', 'Boron', 'Titanium Dioxide'], 'LD50_24h': [219.6, 19.5, None], 'LD50_48h': [7.483, 6.7, None]}
df = pd.DataFrame(data)

plt.bar(df['Nanoparticle'], df['LD50_24h'], color='blue')
plt.title('24h LD50 of Nanoparticles')
plt.ylabel('LD50 (mg L-1)')
plt.show()
```

### Results
The bar chart above illustrates the 24h LD50 values for different nanoparticles, highlighting the relative toxicity of aluminum compared to boron and titanium dioxide.
