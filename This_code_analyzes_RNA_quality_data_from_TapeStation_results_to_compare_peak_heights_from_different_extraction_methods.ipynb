### RNA Quality Analysis from TapeStation Results
This analysis will compare the peak heights of RNA extracted from nuclei using different methods.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load RNA quality data
rna_data = pd.read_csv('tapestation_results.csv')

# Plotting peak heights
plt.figure(figsize=(10,6))
plt.bar(rna_data['Method'], rna_data['PeakHeight'], color='skyblue')
plt.title('Comparison of RNA Peak Heights from Different Extraction Methods')
plt.xlabel('Extraction Method')
plt.ylabel('Peak Height (A.U.)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Discussion
This plot visualizes the differences in RNA peak heights based on the extraction method used, highlighting the effectiveness of each method.

```python
# Save the plot
plt.savefig('rna_peak_heights_comparison.png')
```
