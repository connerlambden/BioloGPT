Step-by-step analysis of gene expression data to study mitochondrial biogenesis markers in lasmiditan-treated versus control PD mice.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('path/to/mitochondrial_markers.csv')
plt.figure(figsize=(8,4))
plt.plot(df['Time'], df['Expression'], marker='o', color='#6A0C76')
plt.title('Mitochondrial Marker Expression Over Time')
plt.xlabel('Time (days)')
plt.ylabel('Expression Level')
plt.grid(True)
plt.show()
```

This analysis helps highlight the temporal changes in mitochondrial biogenesis in response to lasmiditan treatment.
