This notebook will analyze gene expression data from Panc-1 cells to identify transcription factors that may facilitate dedifferentiation.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('panc1_gene_expression.csv')
# Analyze the expression of key transcription factors
transcription_factors = ['CREB', 'MECOM', 'Ptf1a', 'Pdx1']
expression_data = data[transcription_factors]

sns.boxplot(data=expression_data)
plt.title('Expression of Key Transcription Factors in Panc-1 Cells')
plt.show()
```

The boxplot above illustrates the expression levels of key transcription factors in Panc-1 cells, providing insights into potential targets for dedifferentiation.
