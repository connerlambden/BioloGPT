This notebook simulates genetic variation through random assortment during meiosis.

```python
import numpy as np

# Simulate random assortment of alleles
alleles = np.random.choice(['A', 'a'], size=(1000, 2))

# Calculate frequency of each allele
unique, counts = np.unique(alleles, return_counts=True)
frequencies = dict(zip(unique, counts)}
frequencies
```

This code generates allele frequencies based on random assortment.

```python
# Display results
print(frequencies)
```
