# Analyze Th1 differentiation data
# Load necessary libraries and datasets


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load gene expression data
th1_data = pd.read_csv('th1_differentiation_data.csv')

# Analyze the impact of IFN-gamma on Th1 gene expression
ifn_gamma_effect = th1_data[th1_data['cytokine'] == 'IFN-gamma']

# Plot results
plt.figure(figsize=(10,6))
plt.bar(ifn_gamma_effect['gene'], ifn_gamma_effect['expression_level'])
plt.title('Impact of IFN-gamma on Th1 Gene Expression')
plt.xlabel('Genes')
plt.ylabel('Expression Level')
plt.xticks(rotation=45)
plt.show()
```
