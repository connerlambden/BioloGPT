This notebook will analyze RNA-seq data from CFS patients to identify differentially expressed genes associated with immune dysregulation.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load RNA-seq data
rna_data = pd.read_csv('path_to_rna_seq_data.csv')

# Perform differential expression analysis
results = stats.ttest_ind(rna_data['CFS_group'], rna_data['Control_group'])

# Output results
results_df = pd.DataFrame({'Gene': rna_data['Gene'], 'p-value': results.pvalue})
results_df.to_csv('differential_expression_results.csv', index=False)
```
