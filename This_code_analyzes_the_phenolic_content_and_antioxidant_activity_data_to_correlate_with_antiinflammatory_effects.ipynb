### Analysis of Phenolic Compounds and Their Effects
This notebook will analyze the phenolic content and its correlation with the anti-inflammatory effects observed in Lannea velutina and Sorindeia juglandifolia.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for phenolic content and anti-inflammatory activity
data = {
    'Plant': ['Lannea velutina', 'Sorindeia juglandifolia'],
    'Total Phenols (mg GAE/g DE)': [350.1, 463.4],
    'Anti-inflammatory Activity (%)': [47.48, 35.50]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['Plant'], df['Total Phenols (mg GAE/g DE)'], color='blue', alpha=0.6, label='Total Phenols')
plt.bar(df['Plant'], df['Anti-inflammatory Activity (%)'], color='red', alpha=0.6, label='Anti-inflammatory Activity')
plt.title('Phenolic Content and Anti-inflammatory Activity')
plt.ylabel('Values')
plt.legend()
plt.show()
```
