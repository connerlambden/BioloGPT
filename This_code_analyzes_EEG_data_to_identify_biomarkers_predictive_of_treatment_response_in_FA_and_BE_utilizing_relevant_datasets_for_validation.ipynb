### EEG Data Analysis for Predicting Treatment Response in FA and BE

```python
import pandas as pd
import numpy as np

# Load EEG data
# Assuming data is in CSV format
data = pd.read_csv('eeg_data.csv')

# Analyze EEG features
features = data[['theta', 'alpha', 'beta', 'gamma']]

# Calculate correlations with treatment outcomes
correlations = features.corrwith(data['treatment_outcome'])
print(correlations)
```
