# Analysis of Antibacterial Activity of Silver Nanoparticles
This notebook will analyze the antibacterial activity of silver nanoparticles synthesized through mechanochemical and traditional methods.

```python
import pandas as pd

# Load relevant datasets
dataset_mechanochemical = pd.read_csv('mechanochemical_data.csv')
dataset_traditional = pd.read_csv('traditional_data.csv')

# Analyze antibacterial activity
activity_mechanochemical = dataset_mechanochemical['activity'].mean()
activity_traditional = dataset_traditional['activity'].mean()

activity_mechanochemical, activity_traditional
```

## Results
The analysis will provide insights into the comparative antibacterial activity of the two synthesis methods.
