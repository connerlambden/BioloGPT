Este análisis se basa en la composición elemental del cuerpo humano, utilizando datos de estudios previos.

```python
import pandas as pd
# Cargar datos de composición elemental
data = {'Elemento': ['Oxígeno', 'Carbono', 'Hidrógeno', 'Nitrógeno'], 'Porcentaje': [65, 18, 10, 3]}
df = pd.DataFrame(data)
df
```
