This notebook will analyze the recovery rates of proteins purified using various Sephadex grades.

```python
import pandas as pd

# Load the dataset containing recovery rates
recovery_data = pd.read_csv('sephadex_recovery_data.csv')

# Analyze recovery rates by Sephadex grade
recovery_summary = recovery_data.groupby('Sephadex_Grade').mean()
recovery_summary.to_csv('recovery_summary.csv')
```

The analysis results will provide insights into which Sephadex grades yield the highest recovery rates for specific protein sizes.

```python
# Display the summary of recovery rates
print(recovery_summary)
```
