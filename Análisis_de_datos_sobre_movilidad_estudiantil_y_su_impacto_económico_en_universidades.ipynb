Este análisis examina cómo la movilidad estudiantil internacional afecta la economía de las universidades.

```python
import pandas as pd

# Cargar datos sobre movilidad estudiantil
data = pd.read_csv('student_mobility_data.csv')

# Análisis de ingresos generados por estudiantes internacionales
economic_impact = data.groupby('country').agg({'tuition_fees': 'sum', 'living_expenses': 'sum'})

# Mostrar resultados
economic_impact
```

Este análisis proporciona información sobre los ingresos generados por estudiantes internacionales en diferentes países.

```python
# Visualización de los resultados
import matplotlib.pyplot as plt

# Gráfico de ingresos por país
plt.bar(economic_impact.index, economic_impact['tuition_fees'])
plt.title('Ingresos por matrícula de estudiantes internacionales')
plt.xlabel('País')
plt.ylabel('Ingresos (USD)')
plt.show()
```
