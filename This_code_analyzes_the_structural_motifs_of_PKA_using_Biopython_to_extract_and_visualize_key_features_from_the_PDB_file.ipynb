### Step 1: Import necessary libraries and load PDB data.

```python
from Bio import PDB

# Load the PDB structure
pdb_id = '1ATP'
parser = PDB.PDBParser()
structure = parser.get_structure(pdb_id, f'{pdb_id}.pdb')
```

### Step 2: Identify and extract key motifs.

```python
# Extracting motifs
for model in structure:
    for chain in model:
        for residue in chain:
            if residue.get_resname() == 'ASP' and residue.get_id()[1] in [165]:  # DFG motif
                print('DFG Motif found at:', residue)
```
