Load the dataset with measured resolutions for various sample volumes and generate a grouped bar chart using Plotly.

```python
import plotly.graph_objects as go
volumes = ['FSS', 'OB15', 'OB25', 'HPO']
raw = [235.8, 61.7, 91.9, 196.1]
denoised = [96.1, 37.2, 51.0, 103.8]
fig = go.Figure(data=[
    go.Bar(name='Raw', x=volumes, y=raw, marker_color='#D3B9D3'),
    go.Bar(name='Denoised', x=volumes, y=denoised, marker_color='#6A0C76')
])
fig.update_layout(title='Measured Resolution: Raw vs Denoised Volumes', barmode='group',
                  xaxis_title='Volume Type', yaxis_title='Resolution (nm)')
fig.show()
```

This code snippet generates a grouped bar chart to visually compare the measured resolution between raw and denoised image volumes.
