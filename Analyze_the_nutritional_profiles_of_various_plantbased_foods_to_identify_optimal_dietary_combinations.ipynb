This notebook will explore the nutritional data of quinoa, lentils, spinach, and sweet potatoes.

```python
import pandas as pd

# Load nutritional data for analysis
data = pd.read_csv('plant_based_foods_nutrition.csv')

# Display the data
print(data.head())
```

The analysis will focus on key nutrients and their health benefits.

```python
# Analyze key nutrients
nutrient_summary = data.groupby('Food').agg({'Protein': 'sum', 'Fiber': 'sum', 'Vitamins': 'sum'})
print(nutrient_summary)
```
