### Data Analysis of Orbicularis Oculi Training
This notebook analyzes electromyography (EMG) data collected during orbicularis oculi training sessions.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load EMG data
emg_data = pd.read_csv('emg_data.csv')

# Plot EMG activation over time
plt.figure(figsize=(10, 5))
plt.plot(emg_data['Time'], emg_data['Activation'], label='Orbicularis Oculi Activation')
plt.title('EMG Activation of Orbicularis Oculi During Training')
plt.xlabel('Time (s)')
plt.ylabel('Activation Level')
plt.legend()
plt.show()
```
