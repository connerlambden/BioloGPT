This notebook analyzes feeding behavior data of various plankton-eating fish species to identify size preferences and morphological adaptations.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load feeding behavior data
data = pd.read_csv('feeding_behavior_data.csv')

# Analyze size preferences
size_preferences = data.groupby('species')['preferred_prey_size'].mean()

# Visualize results
plt.figure(figsize=(10, 6))
sns.barplot(x=size_preferences.index, y=size_preferences.values)
plt.title('Average Preferred Prey Size by Species')
plt.xlabel('Fish Species')
plt.ylabel('Average Preferred Prey Size (mm)')
plt.show()
```

This analysis provides insights into the size preferences of different fish species, highlighting potential correlations with their morphological traits.
