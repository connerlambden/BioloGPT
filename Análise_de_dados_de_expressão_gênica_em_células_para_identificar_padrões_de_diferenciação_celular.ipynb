Análise de dados de expressão gênica para entender como diferentes tipos de células respondem a estímulos.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('dados_expressao_genica.csv')
sns.boxplot(x='tipo_celula', y='nivel_expressao', data=data)
plt.title('Níveis de Expressão Gênica por Tipo de Célula')
plt.show()
```

Este código analisa os níveis de expressão gênica em diferentes tipos de células, permitindo visualizar como as células se comportam sob diferentes condições.

```python
# Código para análise de dados de expressão gênica
import pandas as pd
import matplotlib.pyplot as plt

dados = pd.read_csv('dados_expressao.csv')
plt.figure(figsize=(10,6))
plt.scatter(dados['gene'], dados['nivel_expressao'])
plt.title('Expressão Gênica em Diferentes Células')
plt.xlabel('Gene')
plt.ylabel('Nível de Expressão')
plt.show()
```

A visualização ajuda a identificar quais genes estão mais ativos em diferentes tipos de células.

```python
# Código para análise de correlação entre genes
import seaborn as sns
import pandas as pd

correlacao = dados.corr()
sns.heatmap(correlacao, annot=True, cmap='coolwarm')
plt.title('Mapa de Calor da Correlação entre Genes')
plt.show()
```
