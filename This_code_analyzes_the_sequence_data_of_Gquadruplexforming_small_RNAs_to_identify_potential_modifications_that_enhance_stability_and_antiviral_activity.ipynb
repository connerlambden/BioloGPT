### Sequence Analysis of G-Quadruplex-Forming Small RNAs
This notebook will analyze the sequences of G-quadruplex-forming small RNAs to identify potential modifications that could enhance their stability and antiviral activity.

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load sequence data
sequences = pd.read_csv('g_quadruplex_sequences.csv')

# Analyze sequences for G-rich content
sequences['G_content'] = sequences['sequence'].apply(lambda x: x.count('G') / len(x))

# Identify potential modifications
sequences['modification'] = np.where(sequences['G_content'] > 0.6, 'Phosphorothioate', 'Standard')

# Display results
sequences.head()
```

### Results
The analysis identifies sequences with high G-content that may benefit from phosphorothioate modifications to enhance stability.

```python
# Save results to a new CSV file
sequences.to_csv('modified_sequences.csv', index=False)
```
