### Analysis of Neurotoxin Detection Methods
This notebook will analyze the detection limits of various neurotoxin detection methods using provided datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

detection_methods = {
    'Method': ['ELISA', 'Immuno-PCR', 'Real-Time PCR', 'Mass Spectrometry', 'Lateral-Flow Assays'],
    'LOD (ng/ml)': [0.5, 0.00000375, 0.1, 0.01, 10]
}
df = pd.DataFrame(detection_methods)

plt.bar(df['Method'], df['LOD (ng/ml)'])
plt.title('Detection Limits of Neurotoxin Detection Methods')
plt.xlabel('Detection Method')
plt.ylabel('Limit of Detection (ng/ml)')
plt.xticks(rotation=45)
plt.show()
```

This analysis provides a visual representation of the detection limits of various neurotoxin detection methods.
