### Data Import and Overview
This block loads the CRC and ALL sample metadata datasets to display sample counts and basic attribute distributions.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets (assuming CSV files available from repository)
crc_data = pd.read_csv('crc_samples.csv')
all_data = pd.read_csv('all_samples.csv')

print('CRC samples:', crc_data.shape[0])
print('ALL samples:', all_data.shape[0])
```

### Visualization of Tumor Stage Distribution in CRC Samples
This section plots the distribution of tumor stages to assess sample grouping.

```python
fig, ax = plt.subplots()
crc_data['tumor_stage'].value_counts().plot(kind='bar', color='#6A0C76', ax=ax)
ax.set_title('CRC Sample Tumor Stages')
ax.set_xlabel('Tumor Stage')
ax.set_ylabel('Counts')
plt.show()
```
