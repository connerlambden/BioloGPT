### Step 1: Load Required Libraries
Import necessary libraries for data analysis and visualization.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2: Load Genetic Data
Load the genetic dataset containing SNP information and phenotypic traits.

```python
data = pd.read_csv('genetic_data.csv')
# Display the first few rows of the dataset
print(data.head())
```

### Step 3: Perform GWAS Analysis
Conduct a genome-wide association study to identify significant SNPs.

```python
# Example of a simple linear regression for GWAS
results = []
for snp in data.columns[1:]:  # Assuming first column is phenotype
    model = sm.OLS(data['phenotype'], data[snp]).fit()
    results.append({'SNP': snp, 'p-value': model.pvalues[0]})
results_df = pd.DataFrame(results)
```
