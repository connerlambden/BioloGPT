### T Cell Response Analysis
This notebook analyzes the ELISPOT assay data to evaluate the effectiveness of the engineered APVs in stimulating T cell responses.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load ELISPOT assay data
data = pd.read_csv('elispot_data.csv')

# Analyze T cell responses
plt.figure(figsize=(10,6))
plt.bar(data['Condition'], data['IFN_gamma_Release'])
plt.title('T Cell IFN-γ Release by Condition')
plt.xlabel('Condition')
plt.ylabel('IFN-γ Release (pg/ml)')
plt.show()
```

### Discussion
This analysis provides insights into the effectiveness of different APV constructs in stimulating T cell responses.
