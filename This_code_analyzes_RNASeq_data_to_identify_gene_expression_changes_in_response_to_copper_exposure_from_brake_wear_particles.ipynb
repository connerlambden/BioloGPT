This notebook will analyze RNA-Seq data from alveolar epithelial cells exposed to brake wear particles.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load RNA-Seq data
data = pd.read_csv('path_to_data.csv')

# Analyze gene expression changes
sns.boxplot(x='condition', y='expression', data=data)
plt.title('Gene Expression Changes in Response to Copper Exposure')
plt.show()
```

The analysis reveals significant changes in gene expression associated with HIF signaling.

```python
# Further analysis of specific genes
hif_genes = data[data['gene'].isin(['HIF1A', 'VEGFA', 'SLC2A1'])]
print(hif_genes)
```
