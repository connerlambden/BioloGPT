### Step 2: Load Hi-C Data
Load your Hi-C contact matrix data for TAD analysis.

```python
import pandas as pd

# Load Hi-C data
hic_data = pd.read_csv('path_to_hic_data.csv')
```

### Step 3: Identify TADs
Use the SpectralTAD package to identify TADs from the Hi-C data.

```python
from SpectralTAD import SpectralTAD

# Initialize and run TAD detection
spectral_tad = SpectralTAD(hic_data)
tads = spectral_tad.detect_tads()
```
