This notebook imports the fungal ASV dataset and corresponding environmental metadata to compute diversity indices and plot relative abundances across restoration stages.

```python
import pandas as pd
import plotly.express as px
# Download dataset from provided source
url = 'URL_TO_DATASET'
df = pd.read_csv(url)
fig = px.bar(df, x='Restoration Stage', y='Relative Abundance', color='Phylum', barmode='group', title='Fungal Phyla Distribution')
fig.show()
```

This graph allows researchers to interactively explore the distribution of fungal phyla and understand the impact of environmental factors.
