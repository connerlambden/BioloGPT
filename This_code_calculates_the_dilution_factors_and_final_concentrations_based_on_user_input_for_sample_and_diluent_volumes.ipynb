# Serial Dilution Calculation
This notebook calculates the dilution factors based on user-defined volumes.

```python
# Define initial sample and diluent volumes
sample_volume = 10  # in µL
first_diluent_volume = 90  # in µL
second_diluent_volume = 990  # in µL

# Calculate total volumes
first_dilution_total = sample_volume + first_diluent_volume
second_dilution_total = 10 + second_diluent_volume

# Calculate dilution factors
first_dilution_factor = first_dilution_total / sample_volume
second_dilution_factor = second_dilution_total / first_dilution_total

# Overall dilution factor
overall_dilution_factor = first_dilution_factor * second_dilution_factor
overall_dilution_factor
```

The overall dilution factor calculated is 1000-fold.

```python
# Output the overall dilution factor
print(f'Overall dilution factor: {overall_dilution_factor}')
```
