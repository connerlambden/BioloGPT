This notebook downloads the clinical dataset for the MRSA bacteremia study, parses key variables, and visualizes outcomes.

```python
import pandas as pd
import plotly.express as px

data = pd.DataFrame([
    {'Group': 'Monotherapy', 'PrimaryOutcome': 24.5, 'TimeToCure': 2.8, 'LengthOfStay': 17.4},
    {'Group': 'Combination', 'PrimaryOutcome': 43, 'TimeToCure': 4.4, 'LengthOfStay': 28.8}
])
fig = px.bar(data, x='Group', y=['PrimaryOutcome','TimeToCure','LengthOfStay'], barmode='group',
             title='Comparison of Key Clinical Outcomes')
fig.show()
```
