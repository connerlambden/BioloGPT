### Data Analysis of Sticky Removal Rates
This notebook will analyze the removal rates of macro and micro stickies based on temperature, disc clearance, and pulp consistency.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for analysis
data = {
    'Temperature (°C)': [70, 80, 90, 100],
    'Removal Rate (%)': [20.1, 21.3, 24.8, 25.7]
}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Temperature (°C)'], df['Removal Rate (%)'])
plt.title('Removal Rates of Stickies at Different Temperatures')
plt.xlabel('Temperature (°C)')
plt.ylabel('Removal Rate (%)')
plt.show()
```

### Discussion
This analysis provides insights into how temperature affects the removal rates of stickies, which is crucial for optimizing recycling processes.
