This notebook loads the experimental dataset including bodyweight, fat deposition, insulin metrics, and oxidative stress markers, then generates grouped bar charts.

```python
import plotly.graph_objs as go
import pandas as pd

data = {
    'Parameter': ['Bodyweight Gain PN21', 'Perigonadal Fat PN21', 'Basal Insulinemia PN14', 'HOMA-β PN21', 'HOMA-IR PN21', 'Glucose Intolerance GTT', 'Islet Area PN21', 'SOD Activity PN21', 'GSH Levels PN21'],
    'Control': [393.9, 0.0850, 0.1513, 44.66, 1.30, 216.4, 8086, 1.381, 1661],
    'MG': [327.5, 0.0695, 0.0983, 17.18, 0.66, 264.5, 5741, 0.835, 791.3]
}
df = pd.DataFrame(data)

fig = go.Figure(data=[go.Bar(name='Control', x=df['Parameter'], y=df['Control']),
                         go.Bar(name='MG', x=df['Parameter'], y=df['MG'])])

fig.update_layout(barmode='group', title='Comparison of Metabolic Parameters', xaxis_tickangle=-45)
fig.show()
print(df)
```

The generated graph clearly contrasts the metabolic parameters between control and MG-exposed groups, illustrating the significant detrimental effects of maternal MG intake.
