### Eye Color Inheritance Simulation
This notebook simulates the inheritance of eye color traits in cheetahs based on the provided genotypes.

```python
import random

def simulate_inheritance(male_genotype, female_genotype, num_offspring):
    offspring = []
    for _ in range(num_offspring):
        offspring.append(random.choice([male_genotype, female_genotype]))
    return offspring

# Simulate 100 offspring
simulate_inheritance('ee', 'Ee', 100)
```
