### Analysis of Random Forest Model Performance
This notebook evaluates the performance metrics (R², RMSE) of Random Forest models predicting woody species diversity using GEDI and Sentinel-2 data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# Assume 'model_performance.csv' contains R2 and RMSE for each diversity index
performance_data = pd.read_csv('model_performance.csv')
```

### Visualization of Model Performance
The following plots illustrate the R² and RMSE values for each diversity index model.

```python
sns.set(style='whitegrid')

# Plot R² values
plt.figure(figsize=(8,6))
ax1 = sns.barplot(x='Diversity Index', y='R2', data=performance_data, palette='Blues_d')
plt.title('Random Forest Model Fit (R²)')
plt.ylabel('R² Value')
plt.xlabel('Diversity Indices')
plt.ylim(0,1)
plt.show()

# Plot RMSE values
plt.figure(figsize=(8,6))
ax2 = sns.barplot(x='Diversity Index', y='RMSE', data=performance_data, palette='Reds_d')
plt.title('Random Forest Model Error (RMSE%)')
plt.ylabel('RMSE Percentage')
plt.xlabel('Diversity Indices')
plt.show()
```

### Summary of Findings
The visualizations reveal that the Species Richness and Shannon Index models exhibit higher R² values and lower RMSE percentages compared to the Simpson Index model, indicating better performance in predicting these diversity indices.
