This notebook analyzes the microbial composition of kefir using quantitative PCR data.

```python
import pandas as pd

# Load the quantitative PCR data
qPCR_data = pd.read_csv('kefir_qPCR_data.csv')

# Analyze the dominant bacterial species
dominant_species = qPCR_data.groupby('Species').sum()

# Display the results
print(dominant_species)
```
