### Data Analysis of Genetic Variation in Dipteran Wings

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('wing_shape_data.csv')
plt.scatter(data['genetic_variation'], data['wing_shape'])
plt.title('Genetic Variation vs Wing Shape')
plt.xlabel('Genetic Variation')
plt.ylabel('Wing Shape')
plt.show()
```
