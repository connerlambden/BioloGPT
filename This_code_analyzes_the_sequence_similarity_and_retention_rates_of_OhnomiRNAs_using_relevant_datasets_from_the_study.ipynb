# Analysis of Ohno-miRNAs
This notebook analyzes the sequence similarity and retention rates of Ohno-miRNAs using datasets from the study.

```python
import pandas as pd

# Load relevant datasets
dataset = pd.read_csv('ohnomiRNAs.csv')

# Analyze sequence similarity
similarity_analysis = dataset['sequence'].apply(lambda x: len(set(x)))

# Display results
similarity_analysis.describe()
```

# Results
The analysis provides insights into the sequence similarity of Ohno-miRNAs.

```python
# Visualize results
import matplotlib.pyplot as plt
plt.hist(similarity_analysis, bins=20)
plt.title('Distribution of Sequence Similarity in Ohno-miRNAs')
plt.xlabel('Sequence Similarity')
plt.ylabel('Frequency')
plt.show()
```
