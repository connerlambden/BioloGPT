### Análisis de Datos de Electroforesis
Este análisis se centra en la separación de proteínas y ácidos nucleicos utilizando datos de electroforesis.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data_electroforesis.csv')
plt.figure(figsize=(10, 6))
plt.bar(data['Tipo'], data['Porcentaje'])
plt.title('Distribución de Tipos de Moléculas en Electroforesis')
plt.xlabel('Tipo de Molécula')
plt.ylabel('Porcentaje')
plt.show()
```

Este gráfico muestra la distribución de tipos de moléculas separadas por electroforesis.
