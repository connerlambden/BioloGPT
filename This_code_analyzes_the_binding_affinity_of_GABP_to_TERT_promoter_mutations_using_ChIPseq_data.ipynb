### Analyze GABP Binding Affinity to TERT Promoter Mutations

```python
import pandas as pd

# Load ChIP-seq data for GABP binding
chip_data = pd.read_csv('gabp_chip_data.csv')

# Analyze binding affinity
binding_affinity = chip_data.groupby('mutation').mean()['affinity']
print(binding_affinity)
```
