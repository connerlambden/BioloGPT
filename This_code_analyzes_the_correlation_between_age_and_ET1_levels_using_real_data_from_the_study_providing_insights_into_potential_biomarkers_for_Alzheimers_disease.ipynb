### Data Analysis of ET-1 Levels and Age in African American Women

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data based on study findings
data = {'Age': [60, 65, 70, 75, 80], 'ET1_Level': [30, 35, 40, 45, 50]}
df = pd.DataFrame(data)

# Calculate correlation
correlation = df['Age'].corr(df['ET1_Level'])
print(f'Correlation between Age and ET-1 Levels: {correlation}')
```

### Visualization of the Correlation

```python
sns.scatterplot(x='Age', y='ET1_Level', data=df)
plt.title('Age vs ET-1 Levels')
plt.xlabel('Age (years)')
plt.ylabel('ET-1 Levels (pg/mL)')
plt.show()
```
