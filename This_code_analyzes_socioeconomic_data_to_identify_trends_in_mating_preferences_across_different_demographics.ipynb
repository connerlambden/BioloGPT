### Socio-Economic Data Analysis for Mating Preferences

```python
import pandas as pd

# Load socio-economic data
socio_economic_data = pd.read_csv('socio_economic_data.csv')

# Analyze trends in mating preferences
trends = socio_economic_data.groupby('income_level')['mating_preference'].value_counts()
```

### Results of the Analysis
This analysis will help identify how income levels influence mating preferences.

```python
trends.plot(kind='bar')
```
