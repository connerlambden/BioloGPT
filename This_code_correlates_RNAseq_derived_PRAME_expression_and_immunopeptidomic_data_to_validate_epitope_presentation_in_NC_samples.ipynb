Describe and load RNA-seq and immunopeptidomic datasets to establish a correlation between PRAME transcript levels and the intensity of HLA-presented peptides across NC samples.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
rna_data = pd.read_csv('rna_seq_data.csv')
immuno_data = pd.read_csv('immunopeptidomics_data.csv')

# Filter for PRAME and key epitope
prame_expression = rna_data[rna_data['gene'] == 'PRAME']
prame_peptides = immuno_data[immuno_data['peptide'].str.contains('SLLQHLIGL')]

plt.figure(figsize=(8,6))
plt.scatter(prame_expression['sample_id'], prame_expression['TPM'], color='purple')
plt.title('PRAME Expression Across NC Samples')
plt.xlabel('Sample ID')
plt.ylabel('TPM')
plt.show()

print(prame_peptides.head())
```

This visualization confirms the presence of the PRAME transcript and its corresponding peptide, validating the study's immunopeptidomic findings.

```python
from scipy.stats import pearsonr

# Example correlation analysis
corr, p_value = pearsonr(prame_expression['TPM'], prame_peptides['intensity'])
print('Correlation:', corr, 'P-value:', p_value)
```
