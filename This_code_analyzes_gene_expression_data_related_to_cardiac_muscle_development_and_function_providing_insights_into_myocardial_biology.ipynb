### Gene Expression Analysis of Cardiac Muscle

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('cardiac_gene_expression.csv')
sns.boxplot(data=data, x='Gene', y='Expression')
plt.title('Gene Expression in Cardiac Muscle')
plt.show()
```
