Below is a concise notebook for reproducing key analyses from DeepCellMap outputs using real data files.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the reconstruction error dataset
errors_df = pd.read_csv('deepcellmap_errors.csv')

# Plot the reconstruction error as a function of misclassification rate
plt.figure(figsize=(8,5))
sns.lineplot(data=errors_df, x='confusion_level', y='association_error', marker='o', color='#6A0C76')
plt.title('Reconstruction Error vs Misclassification Level')
plt.xlabel('Misclassification Level (%)')
plt.ylabel('Association Error (%)')
plt.show()
```

The above code visualizes how error rates change with varying misclassification levels, validating the necessity for error corrections.

```python
clusters_df = pd.read_csv('deepcellmap_clusters.csv')

# Summarize cluster sizes by microglial morphology
cluster_summary = clusters_df.groupby('morphology')['cell_count'].agg(['mean', 'median', 'std']).reset_index()
print(cluster_summary)

# Visualize the distribution of cluster sizes for each morphological type
plt.figure(figsize=(8,5))
sns.boxplot(x='morphology', y='cell_count', data=clusters_df, palette='Set3')
plt.title('Cluster Size Distribution by Microglial Morphology')
plt.xlabel('Morphological Type')
plt.ylabel('Number of Cells per Cluster')
plt.show()
```
