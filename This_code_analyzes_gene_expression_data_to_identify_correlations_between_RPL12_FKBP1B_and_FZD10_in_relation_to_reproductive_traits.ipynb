### Gene Expression Analysis for Reproductive Traits

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_data.csv')
sns.heatmap(data.corr(), annot=True)
plt.title('Correlation between RPL12, FKBP1B, and FZD10')
plt.show()
```
