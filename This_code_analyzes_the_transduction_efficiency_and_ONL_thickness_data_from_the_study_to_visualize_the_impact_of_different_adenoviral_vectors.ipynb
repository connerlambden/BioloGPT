### Analysis of Transduction Efficiency and ONL Thickness
This notebook will analyze the transduction efficiency and ONL thickness data from the study on high-capacity adenoviral vectors.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for ONL thickness
data = {'Vector Type': ['HC-AdV5.F5', 'HC-AdV5.F50', 'Non-treated'],
        'ONL Thickness (µm)': [36, 39, 27]}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Vector Type'], df['ONL Thickness (µm)'], color=['blue', 'orange', 'gray'])
plt.title('Outer Nuclear Layer Thickness by Vector Type')
plt.xlabel('Vector Type')
plt.ylabel('Thickness (µm)')
plt.show()
```
