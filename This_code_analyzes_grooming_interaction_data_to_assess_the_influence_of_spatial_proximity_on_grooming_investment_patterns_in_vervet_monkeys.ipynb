### Data Preparation
Load the grooming interaction dataset and preprocess it for analysis.

```python
import pandas as pd

data = pd.read_csv('grooming_data.csv')
# Preprocess data as needed.
```

### Analysis
Perform statistical analysis to evaluate the relationship between spatial proximity and grooming investment.

```python
import statsmodels.api as sm

# Define the model
model = sm.OLS(data['grooming_investment'], data[['spatial_proximity', 'group_identity']])
results = model.fit()
print(results.summary())
```

### Visualization
Visualize the results to interpret the findings.

```python
import matplotlib.pyplot as plt

plt.scatter(data['spatial_proximity'], data['grooming_investment'])
plt.title('Grooming Investment vs. Spatial Proximity')
plt.xlabel('Spatial Proximity')
plt.ylabel('Grooming Investment')
plt.show()
```
