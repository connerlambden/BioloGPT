```python
import pandas as pd
import plotly.express as px

data = {'Year': [2005, 2010, 2013, 2014],
        'Kappaphycus_Eucheuma': [2444, 5629, 10394, 10992],
        'Saccharina': [4371, 5147, 5942, 7655],
        'Gracilaria': [936, 1696, 3463, 3752],
        'Undaria': [2440, 1537, 2079, 2359],
        'Porphyra': [1287, 1637, 1861, 1806],
        'Sargassum': [86, 78, 152, 175]}

ndf = pd.DataFrame(data)
fig = px.line(ndf, x='Year', y=ndf.columns.drop('Year'), title='Seaweed Production Trends')
fig.show()
```
