### Data Analysis of CO2 Levels in Aquatic Environments

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for CO2 levels over time
co2_data = {'Time (hours)': [0, 1, 2, 3, 4], 'CO2 Level (ppm)': [400, 450, 500, 600, 700]}
df = pd.DataFrame(co2_data)

# Plotting the data
plt.plot(df['Time (hours)'], df['CO2 Level (ppm)'], marker='o')
plt.title('CO2 Levels Over Time in Sealed Test Tube')
plt.xlabel('Time (hours)')
plt.ylabel('CO2 Level (ppm)')
plt.grid()
plt.show()
```

### Discussion
This analysis provides insights into how CO2 levels change over time in a sealed environment with aquatic organisms.
