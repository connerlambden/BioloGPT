### Data Analysis of EcLeuRS Mutants for ncAA Incorporation

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for EcLeuRS mutants
data = {'Mutant': ['EF9', 'B11'], 'Incorporation Efficiency': [0.84, 0.78]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Mutant'], df['Incorporation Efficiency'], color=['blue', 'orange'])
plt.title('Incorporation Efficiency of EcLeuRS Mutants')
plt.xlabel('Mutant')
plt.ylabel('Efficiency')
plt.show()
```
