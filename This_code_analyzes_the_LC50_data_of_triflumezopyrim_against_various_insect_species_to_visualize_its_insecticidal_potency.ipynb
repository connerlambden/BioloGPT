### LC50 Data Analysis for Triflumezopyrim

```python
import pandas as pd
import plotly.express as px

# Data for LC50 values
data = {'Insect Species': ['CPH', 'PLH', 'BPH', 'GLH'], 'LC50 (ppm)': [1, 2.2, 0.5, 4.5]}
df = pd.DataFrame(data)

# Create bar plot
fig = px.bar(df, x='Insect Species', y='LC50 (ppm)', title='LC50 Values of Triflumezopyrim')
fig.show()
```
