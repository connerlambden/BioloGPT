This notebook loads experimental load data and creates interactive Plotly graphs comparing vertical and lateral capacities across different pile types.

```python
import pandas as pd
import plotly.express as px
# Sample dataset representing capacities of different pile designs
data = pd.DataFrame({
    'Pile_Type': ['Conventional', '30RP', '45RP', '60RP', '90RP'],
    'Vertical_Capacity': [100, 130, 140, 150, 200],
    'Lateral_Capacity': [80, 90, 95, 100, 120]
})
fig = px.bar(data, x='Pile_Type', y=['Vertical_Capacity', 'Lateral_Capacity'],
             barmode='group', title='Load Capacity Comparison of Pile Designs')
fig.show()
```

The above visualization allows users to directly compare how different root orientations affect load sharing capacities.
