Load the single-cell RNA-seq dataset from GEO accession GSE275806 and preprocess the data to cluster glandular cells.

```python
import scanpy as sc
data = sc.read_10x_mtx('path_to_data_directory')
sc.pp.filter_cells(data, min_genes=200)
sc.pp.filter_genes(data, min_cells=3)
sc.tl.pca(data)
sc.pp.neighbors(data)
sc.tl.umap(data)
sc.tl.leiden(data, resolution=0.5)
sc.pl.umap(data, color=['FOXA2', 'LIF'])
```

Visualize the UMAP clustering to identify the unique GE_4 subpopulation and quantify its marker expression.

```python
import matplotlib.pyplot as plt
sc.pl.umap(data, color=['LIF'], save='_LIF_expression.png')
```
