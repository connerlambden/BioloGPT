This notebook loads clinical trial data and performs exploratory analysis to compare participant numbers across studies.

```python
import pandas as pd
import plotly.express as px

data = pd.DataFrame([
    {'year': 2022, 'participants': 76, 'study': 'K.C. O’ Donnell'},
    {'year': 2023, 'participants': 25, 'study': 'G. Agin-Liebes'},
    {'year': 2015, 'participants': 12, 'study': 'M.P. Bogenschutz'},
    {'year': 2016, 'participants': 30, 'study': 'P. Glue'},
    {'year': 2014, 'participants': 20, 'study': 'S. Loizaga-Velder'},
    {'year': 2015, 'participants': 15, 'study': 'A. Garcia-Romeu'},
    {'year': 2013, 'participants': 50, 'study': 'G. Thomas'},
    {'year': 2014, 'participants': 15, 'study': 'M.W. Johnson'},
    {'year': 2018, 'participants': 30, 'study': 'E. Dakwar'},
    {'year': 2017, 'participants': 20, 'study': 'M.W. Johnson'},
    {'year': 2018, 'participants': 10, 'study': 'T. Noorani'}
])

fig = px.bar(data, x='year', y='participants', color='study', title='Clinical Trial Participants by Year')
fig.show()
```

The code above employs Plotly to create an interactive bar chart for a clear visual representation of participant data.

```python
import scipy.stats as stats
result = stats.f_oneway(*[group['participants'] for name, group in data.groupby('year')])
print('ANOVA result:', result)
```
