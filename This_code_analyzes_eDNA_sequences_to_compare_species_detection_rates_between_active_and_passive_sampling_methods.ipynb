### eDNA Analysis of Active vs. Passive Sampling
This notebook will analyze eDNA sequences collected from both active and passive sampling methods to compare species detection rates.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load eDNA data
active_data = pd.read_csv('active_sampling_data.csv')
passive_data = pd.read_csv('passive_sampling_data.csv')

# Analyze species detection rates
active_detection = active_data['species'].value_counts()
passive_detection = passive_data['species'].value_counts()

# Plotting the results
plt.figure(figsize=(10,5))
plt.bar(active_detection.index, active_detection.values, alpha=0.5, label='Active Sampling')
plt.bar(passive_detection.index, passive_detection.values, alpha=0.5, label='Passive Sampling')
plt.xlabel('Species')
plt.ylabel('Detection Rate')
plt.title('Comparison of Species Detection Rates')
plt.legend()
plt.show()
```
