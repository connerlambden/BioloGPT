### Analyzing Student Perceptions of AI in Education
This notebook will analyze survey data from medical students regarding their perceptions of AI as a learning tool.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Category': ['Aware of AI', 'Tech Savvy', 'AI Assisted Tasks', 'Believe AI Can Revolutionize Medicine'],
        'Percentage': [84.7, 75.5, 74.2, 90.8]}
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Category'], df['Percentage'], color='purple')
plt.title('Medical Students’ Perceptions of AI')
plt.xlabel('Perception Categories')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.show()
```

### Discussion
This analysis provides insights into how medical students view AI in their education, highlighting areas for potential improvement in teaching methods.
