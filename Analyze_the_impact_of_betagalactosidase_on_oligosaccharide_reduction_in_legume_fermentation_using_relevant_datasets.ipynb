This notebook will analyze the effects of beta-galactosidase on oligosaccharide levels in fermented legumes.

```python
import pandas as pd

# Load relevant datasets
dataset = pd.read_csv('legume_fermentation_data.csv')

# Analyze the impact of beta-galactosidase on oligosaccharide levels
results = dataset.groupby('treatment').mean()
results
```

The analysis shows the average reduction in oligosaccharides across different treatments, highlighting the role of beta-galactosidase.

```python
# Visualize the results
import matplotlib.pyplot as plt

results.plot(kind='bar')
plt.title('Impact of Beta-Galactosidase on Oligosaccharide Levels')
plt.xlabel('Treatment')
plt.ylabel('Average Oligosaccharide Level')
plt.show()
```
