This notebook explores the osmotic behavior of potato cells in different sucrose concentrations, focusing on water movement and cell integrity.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for osmotic behavior
sucrose_concentrations = [0.0, 0.25, 0.5, 0.75]
cell_responses = ['No change', 'Swell', 'Shrink', 'Plasmolysis']

# Create DataFrame
osmotic_data = pd.DataFrame({'Sucrose Concentration (mol/dm³)': sucrose_concentrations, 'Cell Response': cell_responses})

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(osmotic_data['Sucrose Concentration (mol/dm³)'], osmotic_data['Cell Response'])
plt.title('Osmotic Responses of Potato Cells to Sucrose Concentrations')
plt.xlabel('Sucrose Concentration (mol/dm³)')
plt.ylabel('Cell Response')
plt.show()
```

This analysis provides a visual representation of how potato cells respond to different sucrose concentrations, aiding in understanding osmotic effects.
