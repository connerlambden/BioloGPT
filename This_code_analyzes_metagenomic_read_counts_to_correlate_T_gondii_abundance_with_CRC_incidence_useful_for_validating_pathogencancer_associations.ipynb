Downloading and preprocessing metagenomic datasets from PRJEB6070 to quantify T. gondii sequence reads.

```python
import pandas as pd
# Load dataset and filter T_gondii reads
data = pd.read_csv('PRJEB6070_metagenomic_data.csv')
t_gondii = data[data['species'] == 'Toxoplasma gondii']
# Perform Mann-Whitney U test comparing CRC vs non-CRC samples
from scipy.stats import mannwhitneyu
crc_samples = t_gondii[t_gondii['group'] == 'CRC']['read_counts']
non_crc_samples = t_gondii[t_gondii['group'] == 'Non-CRC']['read_counts']
stat, p_value = mannwhitneyu(crc_samples, non_crc_samples)
print('Mann-Whitney U test p-value:', p_value)
```

This analysis quantifies differences in T. gondii abundance and supports the statistical association with CRC risk.
