The notebook downloads the specific demographic data and integrates it with Plotly for visualization.

```python
import plotly.graph_objects as go
import pandas as pd

data = {
    'Population': ['Miramar', 'Mesa de Veinticuatro', 'Otates'],
    'GrowthRate': [1.076, 1.017, 1.042],
    'SeedProduction': [295.3, 931.2, 745]
}
df = pd.DataFrame(data)

fig = go.Figure(data=[
    go.Bar(name='λ (Growth Rate)', x=df['Population'], y=df['GrowthRate'], marker_color='#6A0C76'),
    go.Bar(name='Seed Production', x=df['Population'], y=df['SeedProduction'], marker_color='#FFA500')
])
fig.update_layout(barmode='group', title='Demographic Parameters Across Populations')
fig.show()
```
