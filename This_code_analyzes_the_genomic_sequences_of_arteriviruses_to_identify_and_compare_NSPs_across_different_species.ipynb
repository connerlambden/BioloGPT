### Step 1: Import Required Libraries
Import necessary libraries for genomic analysis.

```python
import pandas as pd
import numpy as np
from Bio import SeqIO
```

### Step 2: Load Genomic Data
Load the genomic sequences of arteriviruses from the NCBI RefSeq database.

```python
# Load genomic data
mammalian_genomes = SeqIO.to_dict(SeqIO.parse('mammalian_arerivirus_genomes.fasta', 'fasta'))
non_mammalian_genomes = SeqIO.to_dict(SeqIO.parse('non_mammalian_arerivirus_genomes.fasta', 'fasta'))
```

### Step 3: Analyze NSPs
Identify and compare NSPs across mammalian and non-mammalian arteriviruses.

```python
# Function to analyze NSPs
def analyze_nsp(genomes):
    nsp_data = []
    for genome in genomes.values():
        nsp_count = genome.seq.count('NSP')  # Example NSP identification
        nsp_data.append({'Genome': genome.id, 'NSP_Count': nsp_count})
    return pd.DataFrame(nsp_data)

mammalian_nsp_analysis = analyze_nsp(mammalian_genomes)
non_mammalian_nsp_analysis = analyze_nsp(non_mammalian_genomes)
```
