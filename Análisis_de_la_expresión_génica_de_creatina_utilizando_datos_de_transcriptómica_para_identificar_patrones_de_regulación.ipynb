Este análisis utilizará datos de expresión génica para investigar cómo los niveles de creatina afectan la regulación de los genes relacionados.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('creatine_expression_data.csv')
sns.boxplot(x='condition', y='expression_level', data=data)
plt.title('Expresión génica de creatina bajo diferentes condiciones')
plt.show()
```

El gráfico anterior muestra cómo la expresión de los genes de creatina varía entre diferentes condiciones experimentales.

```python
# Código para realizar un análisis de correlación entre la expresión de genes y niveles de creatina
genes = data[['gene', 'expression_level']]
correlation = genes.corr()
print(correlation)
```
