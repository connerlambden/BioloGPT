This notebook analyzes absenteeism data from pediatric healthcare workers, focusing on trends before, during, and after the COVID-19 pandemic.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load absenteeism data
absenteeism_data = pd.read_csv('absenteeism_data.csv')

# Analyze trends
absenteeism_trends = absenteeism_data.groupby('year').mean()
plt.plot(absenteeism_trends.index, absenteeism_trends['sick_days'])
plt.title('Absenteeism Trends Over Years')
plt.xlabel('Year')
plt.ylabel('Average Sick Days')
plt.show()
```

This analysis provides insights into how quarantine measures have influenced absenteeism patterns, helping to inform future healthcare policies.
