This notebook will analyze light intensity measurements taken from various canopy layers to determine their effects on plant growth.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load light intensity data
light_data = pd.read_csv('light_intensity_data.csv')

# Plotting light intensity
plt.figure(figsize=(10, 6))
plt.plot(light_data['Height'], light_data['Intensity'], marker='o')
plt.title('Light Intensity Across Canopy Layers')
plt.xlabel('Height (cm)')
plt.ylabel('Light Intensity (μmol m−2 s−1)')
plt.grid()
plt.show()
```

The above plot illustrates the variation in light intensity at different heights within the canopy, providing insights into how light availability changes.
