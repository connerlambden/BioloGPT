# Analyze binding affinities of isoquinuclidines to MOR and KOR


```python
import pandas as pd

# Load docking data
binding_data = pd.read_csv('isoquinuclidine_binding_affinities.csv')

# Analyze binding affinities
binding_summary = binding_data.groupby('compound').agg({'MOR_affinity': 'mean', 'KOR_affinity': 'mean'})

binding_summary.sort_values(by='MOR_affinity', ascending=True)  # Sort by MOR affinity
```

# Summary of binding affinities
This analysis provides insights into which isoquinuclidines have the highest binding affinities for MOR and KOR.

```python
binding_summary.to_csv('binding_summary.csv')  # Save summary to CSV
```
