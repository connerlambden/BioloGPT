This section describes the process of downloading, preprocessing, and visualizing single-cell RNA-seq data from GBM studies to assess T-cell activation changes after GDF15 editing.

```python
import scanpy as sc
adata = sc.read_h5ad('path_to_gbm_dataset.h5ad')
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)
sc.tl.pca(adata, svd_solver='arpack')
sc.pl.pca(adata, color='T_cell_marker')
```

The code above downloads and processes the dataset, filters low-quality cells, performs PCA, and visualizes T-cell marker expression.

```python
sc.pl.umap(adata, color=['T_cell_marker', 'GDF15'], save='_gbm_umap.png')
```
