### Data Analysis of SARS-CoV-2 in Pets
This notebook will analyze sequence data from pets infected with SARS-CoV-2 to identify mutations and potential transmission pathways.

```python
import pandas as pd

# Load sequence data
sequence_data = pd.read_csv('sars_cov_2_pet_sequences.csv')

# Analyze mutations
mutation_analysis = sequence_data['mutation'].value_counts()
```

### Results
The analysis will provide insights into the mutations present in SARS-CoV-2 sequences from pets.

```python
mutation_analysis.plot(kind='bar')
```
