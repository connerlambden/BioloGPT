### Data Analysis of Ophthalmia Neonatorum Incidence and Resistance Patterns

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for analysis
incidence_data = {'Incidence': [2.04], 'Prevalence': [7.79]}
resistance_data = {'Bacteria': ['Staphylococcus aureus', 'Escherichia coli'], 'Resistance to Penicillin': [91.67, 100.00], 'Resistance to Gentamicin': [20.00, 59.30]}

# Create DataFrames
incidence_df = pd.DataFrame(incidence_data)
resistance_df = pd.DataFrame(resistance_data)

# Plotting the incidence and prevalence
plt.figure(figsize=(10, 5))
plt.bar(incidence_df.index, incidence_df['Incidence'], label='Incidence (%)', color='blue')
plt.bar(resistance_df.index + 0.4, resistance_df['Resistance to Penicillin'], label='Resistance to Penicillin (%)', color='red')
plt.title('Ophthalmia Neonatorum Incidence and Antibiotic Resistance')
plt.xticks([0, 1], ['Incidence', 'Resistance to Penicillin'])
plt.ylabel('Percentage (%)')
plt.legend()
plt.show()
```
