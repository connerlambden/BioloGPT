This notebook loads the extracted MDR prevalence data by country and generates an interactive bar chart using Plotly.

```python
import pandas as pd
import plotly.express as px

# Create dataset
data = {'Country': ['Nigeria', 'Ghana', 'Benin', 'Burkina Faso', 'Sierra Leone', 'Mali', 'Gambia', 'Guinea-Bissau', 'Senegal', 'Niger', 'Ivory Coast'],
        'Prevalence': [60, 61.5, 66, 53, 66, 54.5, 55.5, 9, 27.5, 54, 45.5]}
df = pd.DataFrame(data)

# Generate bar chart
fig = px.bar(df, x='Country', y='Prevalence', title='MDR Bacteria Prevalence by Country', color_discrete_sequence=['#6A0C76'])
fig.show()
```

This analysis aids in visualizing inter-country variations in MDR prevalence and supports data-driven recommendations for policy improvements.
