This notebook analyzes seroprevalence data of Toxoplasma gondii in cats, focusing on urban versus rural populations.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for analysis
seroprevalence_data = {
    'Category': ['Total Cats', 'Stray Cats', 'Household Cats'],
    'Percentage': [25.24, 30.77, 17.98]
}

# Create DataFrame
df = pd.DataFrame(seroprevalence_data)

# Plotting
plt.bar(df['Category'], df['Percentage'], color=['blue', 'orange', 'green'])
plt.title('Seroprevalence of Toxoplasma gondii in Cats')
plt.xlabel('Category')
plt.ylabel('Percentage (%)')
plt.show()
```

The analysis reveals significant differences in Toxoplasma gondii infection rates between stray and household cats.
