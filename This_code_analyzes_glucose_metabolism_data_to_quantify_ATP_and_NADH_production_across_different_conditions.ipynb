This notebook will analyze the ATP and NADH production from glucose metabolism under various conditions.

```python
import pandas as pd

def calculate_atp_nadh(glucose_molecules):
    atp_from_glycolysis = 2 * glucose_molecules
    nadh_from_glycolysis = 2 * glucose_molecules
    atp_from_tca = 2 * glucose_molecules
    nadh_from_tca = 8 * glucose_molecules
    total_atp = atp_from_glycolysis + atp_from_tca + (nadh_from_glycolysis * 2.5) + (2 * 1.5)
    return total_atp, nadh_from_glycolysis + nadh_from_tca

calculate_atp_nadh(1) # Example for 1 glucose molecule.
```

This function calculates the total ATP and NADH produced from one glucose molecule.

```python
# Example output
calculate_atp_nadh(1) # Output will show total ATP and NADH for 1 glucose.
```
