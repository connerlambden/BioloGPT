### Data Analysis of Hepatotoxicity in EGFR-TKIs

```python
import pandas as pd

# Load hepatotoxicity data
hepatotoxicity_data = pd.read_csv('hepatotoxicity_data.csv')

# Analyze the data
summary = hepatotoxicity_data.groupby('Drug').agg({'Hepatotoxicity': 'mean'})
summary
```

### Summary of Hepatotoxicity Analysis
This analysis provides insights into the average hepatotoxicity rates associated with gefitinib and afatinib.

```python
summary.to_csv('hepatotoxicity_summary.csv')
```
