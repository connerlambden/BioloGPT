### Gene Expression Analysis for SCNT Studies
This notebook will analyze gene expression data from various SCNT studies to identify genes that are crucial for successful cloning.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('scnt_gene_expression.csv')
sns.boxplot(x='treatment', y='expression_level', data=data)
plt.title('Gene Expression Levels by Treatment')
plt.show()
```
