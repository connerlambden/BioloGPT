Analyze genotype data for PER2 SNPs from Swedish cohorts to correlate allele distributions with depression status.

```python
import pandas as pd
# Load dataset containing PER2 SNP genotypes
df = pd.read_csv('per2_genotype_data.csv')
# Calculate allele frequencies
freqs = df.groupby('SNP')['allele'].value_counts(normalize=True).unstack()
print(freqs)  # Display allele frequency distributions
```

The output will reveal allele frequency differences between depression cases and controls, supporting the association analyses.

```python
import matplotlib.pyplot as plt
freqs.plot(kind='bar', colormap='viridis')
plt.title('Allele Frequencies of PER2 SNPs')
plt.xlabel('SNP')
plt.ylabel('Frequency')
plt.show()
```
