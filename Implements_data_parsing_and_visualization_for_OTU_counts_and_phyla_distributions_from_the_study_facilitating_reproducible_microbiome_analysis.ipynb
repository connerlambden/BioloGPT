This notebook downloads the experimental dataset related to developmental stage OTU counts and prepares it for visualization.

```python
import pandas as pd
import plotly.express as px

data = {
    'Developmental Stage': ['Egg', 'Larval', 'Pupal', 'Adult'],
    'OTU_Count': [254, 426, 289, 251]
}
df = pd.DataFrame(data)
fig = px.bar(df, x='Developmental Stage', y='OTU_Count', title='OTU Counts Across Developmental Stages', color_discrete_sequence=['#6A0C76'])
fig.show()
```

The code above creates an interactive bar chart using Plotly. This is useful for visualizing differences in microbial diversity during development.

This simple analysis provides summary statistics to help interpret the variation in OTU counts.
