This notebook analyzes survey data on hair cutting preferences.

```python
import pandas as pd
# Load survey data
survey_data = pd.read_csv('hair_survey.csv')
# Analyze preferences
preferences = survey_data['preferred_style'].value_counts()
```

The analysis reveals the most popular hair cutting styles among different demographics.

```python
preferences.plot(kind='bar', title='Hair Cutting Style Preferences')
```
