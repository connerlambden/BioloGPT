تحلیل داده‌های رسانایی حرارتی مواد مختلف با استفاده از داده‌های تجربی موجود.

```python
import pandas as pd

# داده‌های رسانایی حرارتی
data = {
    'Material': ['Copper', 'Water', 'Steam', 'Air'],
    'Thermal Conductivity (W/m·K)': [400, 0.6, 0.02, 0.025]
}

# ایجاد DataFrame
thermal_conductivity_df = pd.DataFrame(data)

# نمایش DataFrame
thermal_conductivity_df
```

این کد داده‌های رسانایی حرارتی مواد مختلف را تجزیه و تحلیل می‌کند و به صورت جدول نمایش می‌دهد.

```python
thermal_conductivity_df.plot(kind='bar', x='Material', y='Thermal Conductivity (W/m·K)', title='Thermal Conductivity of Materials')
```
