### RNA-seq Data Analysis for Gene Expression Patterns

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load RNA-seq data
data = pd.read_csv('rna_seq_data.csv')

# Analyze gene expression levels
expression_summary = data.groupby('gene').mean()

# Visualize gene expression patterns
plt.figure(figsize=(10,6))
sns.barplot(x=expression_summary.index, y=expression_summary.values)
plt.title('Average Gene Expression Levels')
plt.xlabel('Genes')
plt.ylabel('Expression Level')
plt.xticks(rotation=90)
plt.show()
```

This analysis provides insights into how nuclear structure influences gene expression.

```python
# Save the results to a file
expression_summary.to_csv('gene_expression_summary.csv')
```
