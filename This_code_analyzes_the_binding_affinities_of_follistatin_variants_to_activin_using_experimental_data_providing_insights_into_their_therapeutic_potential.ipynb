This notebook will analyze the binding affinities of follistatin variants FS-288 and FS-315 to activin, using data from relevant studies.

```python
import pandas as pd

# Data for binding affinities
data = {'Isoform': ['FS-288', 'FS-315'], 'Kd (pM)': [46.5, 432]}
df = pd.DataFrame(data)

# Calculate the difference in binding affinity
binding_difference = df['Kd (pM)'].diff().iloc[-1]

# Display the results
print(df)
print(f'Difference in binding affinity: {binding_difference} pM')
```

The analysis shows that FS-288 has a significantly lower Kd value, indicating a stronger binding affinity to activin compared to FS-315.

```python
# Visualization of binding affinities
import matplotlib.pyplot as plt

plt.bar(df['Isoform'], df['Kd (pM)'], color=['blue', 'orange'])
plt.ylabel('Binding Affinity (Kd in pM)')
plt.title('Binding Affinity of Follistatin Variants to Activin')
plt.show()
```
