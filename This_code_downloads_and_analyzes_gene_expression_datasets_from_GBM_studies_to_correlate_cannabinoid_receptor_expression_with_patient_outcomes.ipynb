Load and preprocess GBM gene expression data along with clinical results.

```python
import pandas as pd
# Download dataset
df = pd.read_csv('https://example.com/gbm_expression.csv')
df.head()
```

Perform differential expression analysis between high and low cannabinoid receptor expressors.

```python
from scipy.stats import ttest_ind
high_expr = df[df['CNR1'] > df['CNR1'].median()]
low_expr = df[df['CNR1'] <= df['CNR1'].median()]
stat, pval = ttest_ind(high_expr['survival'], low_expr['survival'])
print('P-value:', pval)
```

Visualize the survival differences using an interactive box plot.

```python
import plotly.express as px
fig = px.box(df, x='CNR1_group', y='survival', color='CNR1_group', title='Survival vs CNR1 Expression')
fig.show()
```
