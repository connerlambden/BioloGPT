This notebook will guide you through the process of validating metadata entries using interactive error-correction tools.

```python
import pandas as pd

# Load metadata spreadsheet
df = pd.read_excel('metadata.xlsx')

# Define validation rules
validation_rules = {'Field1': ['Value1', 'Value2'], 'Field2': {'type': 'numeric', 'min': 0, 'max': 100}}

# Validate entries
for field, rules in validation_rules.items():
    if isinstance(rules, list):
        df[field] = df[field].apply(lambda x: x in rules)
    else:
        df[field] = df[field].apply(lambda x: rules['min'] <= x <= rules['max'])

# Output validation results
df.to_excel('validated_metadata.xlsx')
```

This code validates metadata entries based on predefined rules and outputs the results to a new spreadsheet.

```python
# Display validation results
validated_df = pd.read_excel('validated_metadata.xlsx')
validated_df.head()
```
