This notebook will analyze genetic diversity datasets from endangered species to evaluate their potential for rapid adaptation.

```python
import pandas as pd

diversity_data = pd.read_csv('genetic_diversity_data.csv')

# Analyze genetic diversity and adaptive potential
results = diversity_data.groupby('species').agg({'heterozygosity': 'mean', 'allelic_richness': 'mean'})
results.to_csv('adaptive_potential_analysis.csv')
```

The results will provide insights into which species have the highest genetic diversity and thus the greatest potential for rapid adaptation.

```python
# Visualize the results
import matplotlib.pyplot as plt

results.plot(kind='bar')
plt.title('Genetic Diversity and Adaptive Potential')
plt.xlabel('Species')
plt.ylabel('Mean Heterozygosity and Allelic Richness')
plt.show()
```
