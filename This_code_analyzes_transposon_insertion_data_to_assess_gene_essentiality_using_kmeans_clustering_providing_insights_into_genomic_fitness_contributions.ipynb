This notebook analyzes transposon insertion data from <em>Mycoplasma pneumoniae</em> to assess gene essentiality.

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

# Load transposon insertion data
data = pd.read_csv('transposon_data.csv')

# K-means clustering for essentiality assessment
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(data[['insertion_count']])

# Add cluster information to the data
data['cluster'] = clusters

# Save the results
data.to_csv('essentiality_results.csv', index=False)
```
