### Bat Immune Gene Expression Analysis
This notebook analyzes immune gene expression data from bats to identify pathways involved in their unique immune responses.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load bat immune gene expression data
data = pd.read_csv('bat_immune_genes.csv')

# Visualize the expression levels of key immune genes
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gene', y='Expression', data=data)
plt.title('Bat Immune Gene Expression Levels')
plt.xticks(rotation=45)
plt.show()
```
