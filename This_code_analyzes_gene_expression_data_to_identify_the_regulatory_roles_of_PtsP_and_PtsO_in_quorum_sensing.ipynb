This notebook will analyze transcriptomic data from <i>Pseudomonas aeruginosa</i> strains with mutations in ptsP and ptsO.

```python
import pandas as pd

# Load the transcriptomic data
data = pd.read_csv('transcriptomic_data.csv')

# Analyze the expression levels of quorum sensing genes
qs_genes = ['lasI', 'phzM', 'hcnA']
qs_expression = data[qs_genes].mean()
qs_expression
```

The analysis reveals the average expression levels of quorum sensing genes across different strains.

```python
qs_expression.plot(kind='bar', title='Average Expression of Quorum Sensing Genes')
```
