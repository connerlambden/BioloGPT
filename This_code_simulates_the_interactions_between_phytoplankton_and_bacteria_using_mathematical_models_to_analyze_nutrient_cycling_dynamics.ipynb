### Simulation of Phytoplankton-Bacteria Interactions
This notebook simulates the interactions based on the mechanisms discussed in the paper.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
overflow_rate = 0.1
mixotrophy_rate = 0.05
exoenzyme_rate = 0.07
ros_detox_rate = 0.03

# Time simulation
time = np.linspace(0, 100, 100)

# Simulate interactions
phytoplankton_growth = np.exp(overflow_rate * time)
bacteria_growth = np.exp(mixotrophy_rate * time)

# Plot results
plt.plot(time, phytoplankton_growth, label='Phytoplankton Growth')
plt.plot(time, bacteria_growth, label='Bacteria Growth')
plt.title('Phytoplankton and Bacteria Growth Over Time')
plt.xlabel('Time')
plt.ylabel('Growth')
plt.legend()
plt.show()
```
