### DNA Weight Calculation
This notebook calculates the weight of DNA in a diploid genome using the average weight of a base pair.

```python
# Constants
num_base_pairs = 6_000_000_000  # Approximate number of base pairs in a diploid human genome
weight_per_base_pair = 650  # Average weight of one base pair in daltons

# Calculate total weight in grams
weight_dna = (num_base_pairs * weight_per_base_pair) / 1_000_000_000_000  # Convert to grams
weight_dna  # Output the weight in grams
```

The calculated weight of DNA in a diploid human genome is approximately 3.9 picograms.

```python
# Output the result
print(f'The weight of DNA in a diploid human genome is approximately {weight_dna:.2f} grams.')
```
