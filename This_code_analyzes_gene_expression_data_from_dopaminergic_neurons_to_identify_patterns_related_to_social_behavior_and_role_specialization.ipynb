### Gene Expression Analysis of Dopaminergic Neurons
This analysis focuses on identifying gene expression patterns in dopaminergic neurons that correlate with social behavior.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('dopaminergic_neurons_expression.csv')
sns.boxplot(x='social_role', y='gene_expression', data=data)
plt.title('Gene Expression by Social Role in Dopaminergic Neurons')
plt.show()
```
