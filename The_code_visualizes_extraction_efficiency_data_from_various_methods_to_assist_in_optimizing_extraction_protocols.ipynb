This notebook downloads the betalain extraction dataset and uses Plotly to generate comparative bar charts for extraction methods.

```python
import plotly.graph_objs as go
methods = ['MW (2 min) + US (10 min)', 'MW (2 min)', 'US (10 min)']
betalain = [51.01, 48.54, 38.44]
antioxidant = [97.44, 128.78, 107.4]

trace1 = go.Bar(x=methods, y=betalain, name='Betalain Content (mg/100g FW)', marker=dict(color='#6A0C76'))
trace2 = go.Bar(x=methods, y=antioxidant, name='Antioxidant Activity (mmol TE/100g FW)', marker=dict(color='#A64AC9'))

fig = go.Figure(data=[trace1, trace2])
fig.update_layout(title='Extraction Method Performance', xaxis_title='Extraction Method', yaxis_title='Measured Values', barmode='group')
fig.show()
```

This code produces an interactive bar plot comparing the effects of different extraction methods on betalain content and antioxidant activity.
