### Gene Expression Analysis of proBDNF and p75 NTR in Inflammation

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load gene expression data
expression_data = pd.read_csv('gene_expression_data.csv')

# Analyze expression levels of proBDNF and p75 NTR
plt.figure(figsize=(10, 6))
plt.bar(expression_data['Gene'], expression_data['Expression_Level'])
plt.title('Gene Expression Levels of proBDNF and p75 NTR')
plt.xlabel('Genes')
plt.ylabel('Expression Level')
plt.show()
```
