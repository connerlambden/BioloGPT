This notebook will analyze RNA-seq data from studies investigating melatonin's effects on immune cells.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load gene expression data
expression_data = pd.read_csv('melatonin_immune_expression.csv')

# Visualize gene expression changes
sns.boxplot(data=expression_data, x='treatment', y='gene_expression')
plt.title('Gene Expression Changes in Response to Melatonin Treatment')
plt.show()
```
