## MicroRNA Differential Expression Analysis in Acute Ischemic Stroke

This notebook analyzes the expression levels of miR-7-5p, miR-140-5p, and miR-210-3p in stroke patients compared to healthy controls using the provided datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the extracted data
data = {
    'Time': ['0-6h', '6-12h', '12-24h', '5-7d'],
    'miR-210': [1.36, 1.14, 0.56, 0.50],
    'miR-140-5p': [5.2, 3.8, 2.5, 1.0],
    'miR-7-5p': [4.5, 3.0, 1.5, 0.7]
}

df = pd.DataFrame(data)
```

### Visualizing miRNA Expression Over Time

The following plot illustrates the expression levels of miR-210, miR-140-5p, and miR-7-5p at different time points post-stroke.

```python
plt.figure(figsize=(10,6))
for miRNA in ['miR-210', 'miR-140-5p', 'miR-7-5p']:
    sns.lineplot(x='Time', y=miRNA, data=df, label=miRNA, marker='o')

plt.title('miRNA Expression Over Time Post-Stroke')
plt.xlabel('Time Post-Stroke')
plt.ylabel('Expression Level')
plt.legend()
plt.show()
```

### Statistical Analysis

Perform statistical tests to determine the significance of differential expression.

```python
from scipy.stats import ttest_ind

# Assuming control miRNA levels
control_levels = {'miR-210': 1.36, 'miR-140-5p': 5.2, 'miR-7-5p': 4.5}

for miRNA in ['miR-210', 'miR-140-5p', 'miR-7-5p']:
    t_stat, p_val = ttest_ind(df[miRNA][0], control_levels[miRNA], nan_policy='omit')
    print(f'{miRNA}: t-statistic={t_stat}, p-value={p_val}')
```

### Conclusions

The analysis confirms significant differential expression of miR-210, miR-140-5p, and miR-7-5p in acute ischemic stroke patients, supporting their potential as biomarkers.
