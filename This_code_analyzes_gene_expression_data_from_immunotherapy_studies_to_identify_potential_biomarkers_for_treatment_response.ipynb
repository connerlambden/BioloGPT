### Gene Expression Analysis for Immunotherapy Response

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_data.csv')
sns.boxplot(x='treatment', y='gene_expression', data=data)
plt.title('Gene Expression by Treatment Group')
plt.show()
```

This analysis visualizes gene expression levels across different treatment groups, helping identify biomarkers associated with immunotherapy efficacy.
