This notebook analyzes survey data to evaluate the relationship between educational programs and students' attitudes towards biotechnology.

```python
import pandas as pd

# Load the survey data
data = pd.read_csv('slovenian_biotech_survey.csv')

# Analyze the data
summary = data.groupby('school_type').agg({'knowledge_score': 'mean', 'attitude_score': 'mean'})
summary
```

The analysis reveals the average knowledge and attitude scores across different school types.

```python
summary.to_csv('biotech_summary.csv')
```
