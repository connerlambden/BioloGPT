This notebook will analyze heart rate data collected during jumping exercises to identify trends and correlations with fitness levels.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load heart rate data
hr_data = pd.read_csv('heart_rate_jumping.csv')

# Analyze data
plt.figure(figsize=(10, 6))
plt.plot(hr_data['Time'], hr_data['HeartRate'], marker='o')
plt.title('Heart Rate Response During Jumping')
plt.xlabel('Time (minutes)')
plt.ylabel('Heart Rate (bpm)')
plt.grid()
plt.show()
```

This analysis visualizes the heart rate response over time during jumping exercises.
