This notebook analyzes blood viscosity data from various studies, focusing on the relationships between hematocrit, plasma volume, and blood pressure.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load blood viscosity data
data = pd.read_csv('blood_viscosity_data.csv')

# Analyze correlations
correlation = data.corr()
sns.heatmap(correlation, annot=True)
plt.title('Correlation Matrix of Blood Parameters')
plt.show()
```

The analysis reveals significant correlations between blood viscosity, hematocrit, and plasma volume, providing insights into the blood equation.

```python
# Further analysis on specific parameters
plt.scatter(data['hematocrit'], data['viscosity'])
plt.xlabel('Hematocrit (%)')
plt.ylabel('Blood Viscosity (mPa.s)')
plt.title('Hematocrit vs Blood Viscosity')
plt.show()
```
