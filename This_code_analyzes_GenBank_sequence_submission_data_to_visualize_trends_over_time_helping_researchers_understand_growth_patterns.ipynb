### GenBank Sequence Submission Analysis
This notebook will analyze the growth of GenBank sequences over time using publicly available datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for GenBank submissions over the years
years = [1982, 1992, 2002, 2012, 2022, 2025]
sequences = [0.001, 0.1, 1, 10, 100, 4700]  # in billions

# Create a DataFrame
data = pd.DataFrame({'Year': years, 'Sequences (Billions)': sequences})

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Sequences (Billions)'], marker='o')
plt.title('Growth of GenBank Sequences Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Sequences (Billions)')
plt.grid()
plt.show()
```

### Discussion
This analysis provides insights into the exponential growth of GenBank, highlighting the impact of technological advancements on genomic data availability.
