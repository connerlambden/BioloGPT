### Protein Sequence Analysis
This section will analyze a given protein sequence to count the number of amino acids and their types.

```python
# Example Python code to analyze protein sequences
from collections import Counter

def analyze_protein_sequence(sequence):
    amino_acid_count = Counter(sequence)
    total_amino_acids = sum(amino_acid_count.values())
    return total_amino_acids, amino_acid_count

# Example usage
sequence = 'MKTAYIAKQRQISFVKSHFSRQD'  # Example protein sequence
analyze_protein_sequence(sequence)  # Returns total count and composition.
```
