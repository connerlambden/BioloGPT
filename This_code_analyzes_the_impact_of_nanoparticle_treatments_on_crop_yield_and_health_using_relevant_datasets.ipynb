### Analysis of Nanoparticle Effects on Crop Yield
This notebook will analyze the effects of different nanoparticle treatments on crop yield and health using data from recent studies.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for nanoparticle treatments
data = {
    'Treatment': ['Control', '5 mg/L ZnONPs', '10 mg/L ZnONPs', '20 mg/L ZnONPs'],
    'Disease Incidence (%)': [0, 26.2, 58.5, 11.8],
    'Mortality (%)': [0, 36.6, 69.0, 10.0]
}
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(df['Treatment'], df['Disease Incidence (%)'], color='blue', alpha=0.6, label='Disease Incidence (%)')
plt.bar(df['Treatment'], df['Mortality (%)'], color='red', alpha=0.6, label='Mortality (%)')
plt.title('Effects of ZnONPs on Disease Incidence and Mortality')
plt.xlabel('Treatment')
plt.ylabel('Percentage (%)')
plt.legend()
plt.show()
```

### Discussion
This analysis provides insights into how different concentrations of ZnONPs affect disease incidence and mortality in crops.
