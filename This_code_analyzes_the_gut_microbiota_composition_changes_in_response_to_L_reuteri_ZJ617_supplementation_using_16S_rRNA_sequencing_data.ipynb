### Analyzing Gut Microbiota Composition
This notebook will analyze the changes in gut microbiota composition in mice supplemented with L. reuteri ZJ617.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load microbiota data
microbiota_data = pd.read_csv('microbiota_data.csv')

# Analyze composition
composition = microbiota_data.groupby('Group').mean()

# Plot results
composition.plot(kind='bar')
plt.title('Gut Microbiota Composition')
plt.ylabel('Relative Abundance')
plt.show()
```
