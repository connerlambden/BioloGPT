Step-by-step: Load forest inventory data from 150 sites and compute deviations in tree species, deadwood, removal, and stand maturity.

```python
import pandas as pd
import numpy as np

data = pd.read_csv('forest_inventory.csv')
# Compute deviations based on natural reference values for each component
data['deviation_species'] = np.abs(data['species_index'] - data['reference_species'])
data['deviation_deadwood'] = np.abs(data['deadwood'] - data['reference_deadwood'])
data['deviation_removal'] = np.abs(data['removed_trees'] - data['reference_removal'])
data['deviation_maturity'] = np.abs(data['maturity'] - data['reference_maturity'])

data['ForMIX'] = data['deviation_species'] + data['deviation_deadwood'] + data['deviation_removal'] + data['deviation_maturity']
print(data[['ForMIX']].describe())
```

The outputs summarize the distribution of ForMIX, enabling further correlation analyses with ecological metrics.

```python
# Additional correlation analysis
import scipy.stats as stats
correlation, p_value = stats.spearmanr(data['ForMIX'], data['biodiversity_index'])
print('Spearman correlation:', correlation, 'p-value:', p_value)
```
