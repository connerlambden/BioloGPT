Detailed description of data preprocessing and statistical analysis using real phosphorus and porina performance datasets.

```python
import pandas as pd
import plotly.express as px

data = pd.read_csv('phosphorus_endophyte_dataset.csv')
fig = px.box(data, x='Endophyte_Status', y='Porina_Weight_Gain', color='Phosphorus_Level', title='Porina Weight Gain vs Endophyte Status and Phosphorus Levels')
fig.show()
```

Step-by-step explanation covering data input, normalization, statistical testing, and interactive visualization generation.

```python
# Additional analysis code
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('Porina_Weight_Gain ~ C(Phosphorus_Level) * C(Endophyte_Status)', data=data).fit()
print(sm.stats.anova_lm(model, typ=2))
```
