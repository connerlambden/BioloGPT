This notebook will analyze the pH-dependent redox activity of pyocyanin using datasets from recent studies.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load relevant datasets
pyocyanin_data = pd.read_csv('pyocyanin_data.csv')

# Analyze the relationship between pH and antimicrobial activity
plt.figure(figsize=(10,6))
plt.scatter(pyocyanin_data['pH'], pyocyanin_data['antimicrobial_activity'])
plt.title('pH vs Antimicrobial Activity of Pyocyanin')
plt.xlabel('pH')
plt.ylabel('Antimicrobial Activity')
plt.grid()
plt.show()
```

This analysis will help identify the optimal pH range for enhancing pyocyanin's antimicrobial properties.
