## Correlation Analysis of Temperature and Color Proportions in Pachnoda iskuulka

This analysis evaluates the relationship between rearing temperature and the proportions of black, yellow, and red coloration in Pachnoda iskuulka beetles.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Load the dataset
data = pd.DataFrame({
    'Temperature': [24, 27, 30],
    'Black': [64.4, 64.4, 63.24],
    'Yellow': [19.39, 19.39, 20.16],
    'Red': [14.46, 14.46, 16.60],
    'Development_Time': [101, 88, 77]
})

data.head()
```

### Visualization of Color Proportions Across Temperatures

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Temperature', y='value', hue='color', data=pd.melt(data, ['Temperature'], value_vars=['Black', 'Yellow', 'Red'], var_name='color', value_name='value'))
plt.title('Color Proportions at Different Temperatures')
plt.xlabel('Temperature (°C)')
plt.ylabel('Coverage (%)')
plt.legend(title='Color')
plt.show()
```

### Correlation Between Temperature and Black Coloration

```python
black_corr, black_p = pearsonr(data['Temperature'], data['Black'])
print(f'Pearson correlation between Temperature and Black Coverage: {black_corr:.2f} (p-value: {black_p:.4f})')
```

### Interpretation
A negative correlation indicates that as temperature increases, the proportion of black coloration decreases.
