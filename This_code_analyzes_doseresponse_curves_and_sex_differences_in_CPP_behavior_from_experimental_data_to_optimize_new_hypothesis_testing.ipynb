Detailed exploration of the experimental dataset including exposure levels, CPP scores, and sex differences.

```python
import pandas as pd
import plotly.express as px

# Download dataset from relevant source
# df = pd.read_csv('experimental_data.csv')
# For demonstration, generating synthetic data
import numpy as np
np.random.seed(42)

doses = ['Low', 'Medium', 'High']
sex = ['Male', 'Female']
data = []
for d in doses:
    for s in sex:
        score = np.random.normal(loc=10 if d=='High' else (5 if d=='Medium' else 0), scale=2, size=10)
        for sc in score:
            data.append({'Dose': d, 'Sex': s, 'CPP_Score': sc})

 df = pd.DataFrame(data)
fig = px.box(df, x='Dose', y='CPP_Score', color='Sex', title='CPP Score Distribution by Dose and Sex')
fig.show()
```

The plot above visualizes the CPP score distributions across different THC doses and between sexes, providing insights for further experimental design.

```python
# Additional code for statistical analysis could be added here
import statsmodels.api as sm
import statsmodels.formula.api as smf

model = smf.ols('CPP_Score ~ C(Dose) * C(Sex)', data=df).fit()
print(model.summary())
```
