### BRAT1 Expression Analysis in Glioblastoma

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load expression data
expression_data = pd.read_csv('brat1_expression_data.csv')

# Analyze correlation with survival
survival_data = pd.read_csv('patient_survival_data.csv')

# Merge datasets
merged_data = pd.merge(expression_data, survival_data, on='patient_id')

# Plotting
plt.scatter(merged_data['brat1_expression'], merged_data['survival_time'])
plt.title('BRAT1 Expression vs Survival Time')
plt.xlabel('BRAT1 Expression Level')
plt.ylabel('Survival Time (months)')
plt.show()
```

This analysis provides insights into how BRAT1 expression levels correlate with patient survival, potentially guiding therapeutic strategies.
