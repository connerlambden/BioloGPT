### Analyzing the Impact of Solvent Polarity on Enzyme Activity

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load experimental data on enzyme activity in different solvents
data = {'Solvent': ['Hexane', 'Ethanol', 'Methanol'], 'Activity': [73.8, 65.5, 19.4]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Solvent'], df['Activity'], color=['#ff9999','#66b3ff','#99ff99'])
plt.title('Enzyme Activity in Different Solvents')
plt.xlabel('Solvent')
plt.ylabel('Activity (%)')
plt.show()
```

This analysis visualizes how different solvents affect enzyme activity, highlighting the importance of solvent choice in catalytic processes.
