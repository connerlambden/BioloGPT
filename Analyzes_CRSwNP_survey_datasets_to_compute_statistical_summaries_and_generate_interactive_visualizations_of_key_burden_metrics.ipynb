This notebook loads and processes the CRSwNP survey dataset to produce descriptive statistics and visualizations for sleep quality and economic burden.

```python
import pandas as pd
import plotly.express as px

# Load the CRSwNP survey data
data = pd.read_csv('CRSwNPs_data.csv')
stats = data.describe()

# Create a histogram for sleep quality
fig = px.histogram(data, x='sleep_quality', title='Distribution of Sleep Quality Scores in CRSwNP Patients')
fig.show()

# Print statistical summary
print(stats)
```

This analysis helps in understanding the correlation between sleep disturbances and reported quality of life outcomes.
