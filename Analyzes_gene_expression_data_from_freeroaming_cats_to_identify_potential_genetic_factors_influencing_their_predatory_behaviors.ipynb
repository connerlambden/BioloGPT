### Analysis of Gene Expression in Free-Roaming Cats
This notebook analyzes miRNA expression profiles to identify genetic factors associated with predatory behaviors in free-roaming cats.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('path_to_mirna_expression_data.csv')
```

#### Data Exploration
We begin by exploring the distribution of miRNA expression levels across different populations of free-roaming cats.

```python
sns.boxplot(x='Population', y='miRNA_expression', data=data)
plt.title('miRNA Expression by Population')
plt.show()
```

#### Differential Expression Analysis
Identify miRNAs that are significantly upregulated or downregulated in predatory versus non-predatory cats.

```python
# Example using a simple threshold for differential expression
predatory = data[data['Behavior'] == 'Predatory']['miRNA_expression']
non_predatory = data[data['Behavior'] == 'Non-Predatory']['miRNA_expression']

diff = predatory.mean() - non_predatory.mean()
print(f'Differential Expression: {diff}')
```
