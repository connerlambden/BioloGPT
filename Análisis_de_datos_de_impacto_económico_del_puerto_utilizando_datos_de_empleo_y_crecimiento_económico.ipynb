Análisis de datos económicos relacionados con el Nuevo Puerto de Amador.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de ejemplo
data = {'Año': [2020, 2021, 2022, 2023, 2024], 'Empleo': [1000, 2000, 3000, 4000, 5000]}
df = pd.DataFrame(data)

# Gráfico
plt.plot(df['Año'], df['Empleo'])
plt.title('Proyección de Empleo en el Nuevo Puerto de Amador')
plt.xlabel('Año')
plt.ylabel('Empleo')
plt.show()
```

Este análisis muestra la proyección de empleo en el Nuevo Puerto de Amador a lo largo de los años.
