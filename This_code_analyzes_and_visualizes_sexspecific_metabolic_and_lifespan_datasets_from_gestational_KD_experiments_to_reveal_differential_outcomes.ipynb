We load the gestational KD dataset and create visualizations for lifespan and sex ratio differences.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Diet': ['Control', 'Ketogenic'], 'Lifespan': [661, 512], 'Female_Percent': [52.9, 39.6]}
df = pd.DataFrame(data)
plt.figure(figsize=(8,4))
plt.bar(df['Diet'], df['Lifespan'], color=['#6A0C76', '#FF5733'])
plt.xlabel('Diet')
plt.ylabel('Lifespan (days)')
plt.title('Lifespan by Gestational Diet')
plt.show()
```
