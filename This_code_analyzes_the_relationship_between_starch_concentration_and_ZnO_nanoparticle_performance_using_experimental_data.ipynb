### Data Analysis of Starch Concentration Effects on ZnO Nanoparticles

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
starch_concentration = [1, 2, 3, 4, 5]
photocatalytic_performance = [15, 30, 45, 60, 50]  # hypothetical values

# Create DataFrame
data = pd.DataFrame({'Starch Concentration': starch_concentration, 'Performance': photocatalytic_performance})

# Plotting
plt.plot(data['Starch Concentration'], data['Performance'], marker='o')
plt.title('Effect of Starch Concentration on ZnO Performance')
plt.xlabel('Starch Concentration (g/L)')
plt.ylabel('Photocatalytic Performance (degradation rate)')
plt.grid()
plt.show()
```

This analysis provides insights into how varying starch concentrations impact the photocatalytic performance of ZnO nanoparticles.
