### Data Analysis of Lemon Microbiome
This notebook analyzes the microbial diversity in lemon samples using sequencing data.

```python
import pandas as pd

# Load the sequencing data
bacterial_data = pd.read_csv('bacterial_sequences.csv')
fungal_data = pd.read_csv('fungal_sequences.csv')

# Analyze diversity
bacterial_diversity = bacterial_data['diversity_index'].mean()
fungal_diversity = fungal_data['diversity_index'].mean()

print(f'Bacterial Diversity: {bacterial_diversity}')
print(f'Fungal Diversity: {fungal_diversity}')
```
