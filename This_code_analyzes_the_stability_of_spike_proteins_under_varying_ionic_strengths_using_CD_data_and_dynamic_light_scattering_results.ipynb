Analyze CD spectra data to determine secondary structure content at different ionic strengths.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CD data
cd_data = pd.read_csv('cd_spectra.csv')

# Plotting CD spectra for different ionic strengths
plt.figure(figsize=(10, 6))
for ionic_strength in cd_data['ionic_strength'].unique():
    subset = cd_data[cd_data['ionic_strength'] == ionic_strength]
    plt.plot(subset['wavelength'], subset['ellipticity'], label=f'Ionic Strength: {ionic_strength} mM')
plt.title('CD Spectra of Spike Proteins at Varying Ionic Strengths')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Ellipticity (mdeg)')
plt.legend()
plt.show()
```
