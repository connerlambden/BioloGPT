### Step 1: Import Libraries
Import necessary libraries for data analysis and visualization.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset containing organoid morphological data
dataset = pd.read_csv('organoid_data.csv')

# Display the first few rows of the dataset
dataset.head()
```

### Step 2: Analyze Correlation
Calculate the correlation between Feret diameter and quality indicators.

```python
# Calculate correlation
correlation = dataset['Feret_diameter'].corr(dataset['Quality'])
print(f'Correlation between Feret diameter and quality: {correlation}')
```

### Step 3: Visualize Results
Create a scatter plot to visualize the relationship.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dataset, x='Feret_diameter', y='Quality')
plt.title('Feret Diameter vs. Organoid Quality')
plt.xlabel('Feret Diameter (μm)')
plt.ylabel('Quality Score')
plt.show()
```
