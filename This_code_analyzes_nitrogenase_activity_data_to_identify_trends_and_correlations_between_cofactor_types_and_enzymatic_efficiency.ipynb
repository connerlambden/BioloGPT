### Analyzing Nitrogenase Activity Data
This notebook will analyze the nitrogenase activity data from various strains of <i>Azotobacter vinelandii</i>.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Strain': ['DJ2560', 'DJ2821', 'DJ2831'],
    'N2 Activity': [161, 0, 141],
    'C2H2 Activity': [536, 183, 472]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 5))
plt.bar(df['Strain'], df['N2 Activity'], color='blue', label='N2 Activity')
plt.bar(df['Strain'], df['C2H2 Activity'], color='orange', label='C2H2 Activity', alpha=0.7)
plt.title('Nitrogenase Activity Comparison')
plt.xlabel('Strain')
plt.ylabel('Activity (nmol/min/mg)')
plt.legend()
plt.show()
```

This analysis provides insights into the differences in nitrogenase activity among strains with varying cofactor compositions.
