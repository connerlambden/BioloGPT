Import necessary libraries and load datasets for envirotyping, yield, and stay-green traits.

```python
import pandas as pd
import plotly.express as px

data = pd.read_csv('envirotyping_yield_data.csv')
fig = px.scatter(data, x='water_stress', y='yield', color='stay_green', title='Yield vs Water Stress')
fig.show()
```

Explain the correlation analysis between yield and stay-green metrics across diverse environments.

```python
import numpy as np
correlations = data[['yield','stay_green']].corr()
print(correlations)
```
