### Allele Frequency Calculation
This notebook calculates the frequencies of the HbS and wild-type alleles based on provided genotype data.

```python
# Sample genotype data
# Genotype: 0 = HbAA, 1 = HbAS, 2 = HbSS
genotypes = [0, 1, 1, 2, 0, 1, 0, 2, 1, 0]  # Example data

# Calculate frequencies
num_hbss = genotypes.count(2)
num_hbas = genotypes.count(1)
num_hbaa = genotypes.count(0)

# Total individuals
n = len(genotypes)

# Frequency calculations
q = num_hbss / n  # Frequency of HbS allele
p = (num_hbaa + 0.5 * num_hbas) / n  # Frequency of wild-type allele

q, p
```

### Results
The calculated frequencies of the HbS and wild-type alleles will provide insights into the genetic landscape of the population.

```python
# Display results
print(f'Frequency of HbS allele (q): {q}')
print(f'Frequency of wild-type allele (p): {p}')
```
