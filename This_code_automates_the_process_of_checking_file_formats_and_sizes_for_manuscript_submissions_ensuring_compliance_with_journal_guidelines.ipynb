# File Submission Compliance Checker
This notebook checks if the attached files meet the journal's submission guidelines.

```python
def check_file_compliance(files):
    accepted_formats = ['.doc', '.docx', '.pdf', '.rtf', '.jpg', '.png']
    max_size_mb = 10  # Example size limit
    for file in files:
        if not any(file.endswith(ext) for ext in accepted_formats):
            print(f"{file} is not an accepted format.")
        elif os.path.getsize(file) > max_size_mb * 1024 * 1024:
            print(f"{file} exceeds the size limit.")
        else:
            print(f"{file} is compliant.")
```

## Usage
Run the function with a list of file names to check compliance.

```python
check_file_compliance(['manuscript.docx', 'figure1.jpg', 'data.csv'])
```
