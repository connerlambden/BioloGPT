This notebook will analyze RNA-seq data to identify genes regulated by histamine exposure.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load RNA-seq data
rna_data = pd.read_csv('path_to_rna_seq_data.csv')

# Analyze gene expression changes
sns.boxplot(x='condition', y='expression', data=rna_data)
plt.title('Gene Expression Changes in Response to Histamine')
plt.show()
```
