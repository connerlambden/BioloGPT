This notebook extracts real data on gene expressions and stimulation parameters from the study, preparing it for analysis.

```python
import pandas as pd
# Downloading and reading the dataset containing electrostimulation parameters and gene expression levels
data = pd.read_csv('endothelial_data.csv')
df = pd.DataFrame(data)
print(df.head())
```

Next, the code cleans the data, performs correlation analysis between PEDOT current and nitric oxide production, and visualizes the trends.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(df['PEDOT_current'], df['NO_production'], color='tomato')
plt.title('Correlation between PEDOT Current and NO Production')
plt.xlabel('PEDOT Current (mA)')
plt.ylabel('NO Production (arbitrary units)')
plt.show()
```

The analysis confirms trends observed in the experimental data, supporting enhanced endothelial responses with PEDOT electrodes.

```python
# Final summary of correlation analysis
correlation = df['PEDOT_current'].corr(df['NO_production'])
print('Correlation coefficient:', correlation)
```
