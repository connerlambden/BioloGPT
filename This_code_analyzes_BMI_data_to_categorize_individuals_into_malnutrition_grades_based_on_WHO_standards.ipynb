### BMI Classification Analysis
This notebook will analyze BMI data and categorize individuals into malnutrition grades according to WHO standards.

```python
import pandas as pd

def classify_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal weight'
    elif 25 <= bmi < 30:
        return 'Overweight'
    elif 30 <= bmi < 35:
        return 'Obesity Class I'
    elif 35 <= bmi < 40:
        return 'Obesity Class II'
    else:
        return 'Obesity Class III'

# Example DataFrame
bmi_data = pd.DataFrame({'BMI': [17.5, 22.0, 27.5, 32.0, 40.5]})
bmi_data['Classification'] = bmi_data['BMI'].apply(classify_bmi)
print(bmi_data)
```
