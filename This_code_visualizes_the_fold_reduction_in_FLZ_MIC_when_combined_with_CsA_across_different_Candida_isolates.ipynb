Load the extracted MIC data to calculate and visualize the reduction factor in FLZ MIC with the combination therapy.

```python
import pandas as pd
import plotly.express as px

data = [
    {'isolate': 'C. albicans ATCC 64124', 'MIC_FLZ_monotherapy': 64, 'MIC_FLZ_combination': 0.25},
    {'isolate': 'C. glabrata UPV 15-202', 'MIC_FLZ_monotherapy': 32, 'MIC_FLZ_combination': 0.25},
    {'isolate': 'C. parapsilosis ATCC 22019', 'MIC_FLZ_monotherapy': 1, 'MIC_FLZ_combination': 0.25}
]
df = pd.DataFrame(data)
df['MIC_Reduction'] = df['MIC_FLZ_monotherapy'] / df['MIC_FLZ_combination']
fig = px.bar(df, x='isolate', y='MIC_Reduction', title='Fold Reduction in FLZ MIC with CsA Combination', labels={'MIC_Reduction': 'Reduction Factor'})
fig.show()
```

The bar chart portrays the dramatic reduction in FLZ MIC when used with CsA, emphasizing the potential of the combination approach to mitigate antifungal resistance.
