### Step 1: Import necessary libraries and load datasets.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load nitrogen use efficiency data
nue_data = pd.read_csv('nue_data.csv')
```

### Step 2: Analyze the correlation between early biomass and grain yield.

```python
correlation = nue_data['early_biomass'].corr(nue_data['grain_yield'])
print(f'Correlation between early biomass and grain yield: {correlation}')
```

### Step 3: Visualize the relationship.

```python
plt.scatter(nue_data['early_biomass'], nue_data['grain_yield'])
plt.title('Early Biomass vs Grain Yield')
plt.xlabel('Early Biomass (kg/ha)')
plt.ylabel('Grain Yield (t/ha)')
plt.show()
```
