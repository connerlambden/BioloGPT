This notebook will analyze the correlation between morphological characteristics and phenotypic traits using datasets from Cinnamomum camphora.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
dataset = pd.read_csv('cinnamomum_data.csv')

# Analyze correlation
correlation = dataset.corr()
sns.heatmap(correlation, annot=True)
plt.title('Correlation between Morphological and Phenotypic Traits')
plt.show()
```

The heatmap above illustrates the correlation between various morphological and phenotypic traits, providing insights into their relationships.
