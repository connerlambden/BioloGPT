### Data Preparation
Load datasets containing cognitive assessments and noise exposure measurements.

```python
import pandas as pd

# Load cognitive assessment data
df_cognitive = pd.read_csv('cognitive_assessment.csv')

# Load noise exposure data
df_noise = pd.read_csv('noise_exposure.csv')
```

### Data Analysis
Merge datasets and analyze the correlation between noise exposure and cognitive scores.

```python
# Merge datasets on participant ID
df_merged = pd.merge(df_cognitive, df_noise, on='participant_id')

# Analyze correlation
correlation = df_merged.corr()['cognitive_score']['noise_exposure']
print(f'Correlation between noise exposure and cognitive score: {correlation}')
```
