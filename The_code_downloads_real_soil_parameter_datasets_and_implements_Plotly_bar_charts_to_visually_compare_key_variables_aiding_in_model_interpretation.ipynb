This notebook downloads the soil parameter dataset and generates a Plotly bar chart comparing key parameters such as permeability, water content, and others.

```python
import plotly.graph_objects as go

parameters = ['Saturated Permeability (m/s)', 'Volumetric Water Content', 'Air-entry Value (kPa)', "Effective Cohesion (kPa)", 'Friction Angle (°)', 'Saturated Unit Weight (kN/m^3)']
values = [3.13e-8, 0.523, 204, 5, 29, 16.5]

fig = go.Figure([go.Bar(x=parameters, y=values, marker_color='#6A0C76')])
fig.update_layout(title='Extracted Soil Parameters', xaxis_title='Parameter', yaxis_title='Value')
fig.show()
```

This code provides a graphical representation of the extracted soil parameters which are crucial for interpreting the model's outputs.
