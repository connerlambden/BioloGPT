This notebook will simulate genetic drift in a population using historical allele frequency data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate genetic drift
population_size = 100
num_generations = 50
allele_frequencies = np.zeros(num_generations)
allele_frequencies[0] = 0.5  # Initial frequency of allele A

for generation in range(1, num_generations):
    allele_frequencies[generation] = np.random.binomial(population_size, allele_frequencies[generation-1]) / population_size

# Plotting the results
plt.plot(allele_frequencies)
plt.title('Genetic Drift Simulation')
plt.xlabel('Generations')
plt.ylabel('Allele Frequency')
plt.show()
```

This simulation demonstrates how allele frequencies can fluctuate over generations due to genetic drift.
