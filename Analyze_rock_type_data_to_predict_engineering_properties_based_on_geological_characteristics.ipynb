This notebook will analyze datasets related to rock types and their engineering properties, focusing on compressive strength and porosity.

```python
import pandas as pd

# Load rock type data
data = pd.read_csv('rock_properties.csv')

# Analyze compressive strength vs porosity
correlation = data['compressive_strength'].corr(data['porosity'])
print(f'Correlation between compressive strength and porosity: {correlation}')
```

The analysis will provide insights into how different rock types can be optimized for engineering applications.

```python
# Visualize the relationship
import matplotlib.pyplot as plt
plt.scatter(data['porosity'], data['compressive_strength'])
plt.title('Compressive Strength vs Porosity')
plt.xlabel('Porosity')
plt.ylabel('Compressive Strength')
plt.show()
```
