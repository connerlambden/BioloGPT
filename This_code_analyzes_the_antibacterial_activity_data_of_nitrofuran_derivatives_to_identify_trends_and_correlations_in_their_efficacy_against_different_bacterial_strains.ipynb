### Antibacterial Activity Analysis of Nitrofuran Derivatives

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for analysis
data = {
    'Compound': ['6-(methylsulfonyl)-8-(4-methyl-4H-1,2,4-triazol-3-yl)-2-(5-nitro-2-furoyl)-2,6-diazaspiro[3.4]octane', '2b', '5c'],
    'MIC_Mtb': [0.0124, 100, 6.2],
    'MIC_S_aureus': [100, 0.171, 0.19]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.bar(df['Compound'], df['MIC_Mtb'], color='blue', label='MIC against M. tuberculosis')
plt.bar(df['Compound'], df['MIC_S_aureus'], color='red', label='MIC against S. aureus', alpha=0.5)
plt.ylabel('MIC (µg/mL)')
plt.title('Antibacterial Activity of Nitrofuran Derivatives')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
