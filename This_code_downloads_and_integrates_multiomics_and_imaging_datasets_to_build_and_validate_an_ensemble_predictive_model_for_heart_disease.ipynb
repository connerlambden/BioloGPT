This section describes the detailed process of loading multi-omics and imaging data followed by feature extraction and model ensemble building.

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
# Load imaging and omics datasets
omics_data = pd.read_csv('omics_data.csv')
imaging_data = pd.read_csv('imaging_data.csv')
# Merge datasets based on patient ID
merged_data = pd.merge(omics_data, imaging_data, on='patient_id')
# Feature selection and training
features = merged_data.drop('label', axis=1)
target = merged_data['label']
model = XGBClassifier()
model.fit(features, target)
print('Model trained with accuracy:', model.score(features, target))
```

The above code snippet demonstrates how to integrate data from different modalities and train an ensemble model to predict heart disease outcomes.

```python
# Further analysis and SHAP explanations can be added here to interpret the model
import shap
explainer = shap.Explainer(model, features)
shap_values = explainer(features)
shap.plots.bar(shap_values)
```
