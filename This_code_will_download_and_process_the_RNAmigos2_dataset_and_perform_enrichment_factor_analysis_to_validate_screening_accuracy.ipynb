Start by downloading the dataset from Zenodo using the provided link and load it into a pandas DataFrame for analysis.

```python
import pandas as pd
import numpy as np

# Replace 'your_dataset.csv' with the actual file path from Zenodo
# Dataset should include compound scores and labels (active/decoy)
data = pd.read_csv('your_dataset.csv')

# Compute enrichment factor at top 1%
top_1pct = int(0.01 * len(data))
data_sorted = data.sort_values(by='score', ascending=False)
enrichment_factor = (data_sorted.head(top_1pct)['active'].sum() / top_1pct) / (data['active'].sum() / len(data))
print('Enrichment Factor at 1%:', enrichment_factor)
```

Visualize the score distribution for active versus decoy compounds using Plotly for an interactive graph.

```python
import plotly.express as px

fig = px.histogram(data, x='score', color='active', barmode='overlay', 
                   title='Distribution of Compound Scores for Active vs Decoy')
fig.update_layout(bargap=0.1, template='plotly_white')
fig.show()
```
