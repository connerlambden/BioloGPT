This notebook analyzes the performance of various encapsulated adhesives based on recent studies.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Adhesive': ['PUR', 'PUR-10', 'PUR-15'],
        '5 Min RT': [0.06, 0.47, 0.41],
        '5 Min COOL': [0.6, 5.23, 1.83],
        '7 Days RT': [12.48, 8.59, 8.05]}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['Adhesive'], df['5 Min RT'], label='5 Min RT')
plt.bar(df['Adhesive'], df['5 Min COOL'], label='5 Min COOL', bottom=df['5 Min RT'])
plt.bar(df['Adhesive'], df['7 Days RT'], label='7 Days RT', bottom=df['5 Min RT'] + df['5 Min COOL'])
plt.title('Adhesive Performance')
plt.xlabel('Adhesive Type')
plt.ylabel('Performance (N/mm)')
plt.legend()
plt.show()
```

This analysis provides insights into the performance of different adhesive formulations.
