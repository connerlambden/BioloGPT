## Cadmium Exposure Gene Expression Analysis
This notebook analyzes the changes in gene expression in the olfactory epithelium of zebrafish following cadmium exposure.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('zebrafish_cadmium_gene_expression.csv')
```

### Differential Gene Expression
Identify genes significantly upregulated or downregulated after cadmium exposure.

```python
diff_expr = data[(data['p_value'] < 0.05) & (abs(data['log2_fold_change']) > 1)]
sns.scatterplot(data=data, x='log2_fold_change', y='-log10(p_value)', hue='regulated')
plt.title('Differential Gene Expression Post-Cadmium Exposure')
plt.xlabel('Log2 Fold Change')
plt.ylabel('-Log10(p-value)')
plt.show()
```

### Pathway Enrichment Analysis
Determine which biological pathways are affected by cadmium exposure.

```python
# Example code for pathway analysis (details depend on available tools and data)
import gseapy as gp
enrich_res = gp.enrichr(gene_list=diff_expr['gene_symbol'].tolist(),
                          gene_sets='KEGG_2021_Human',
                          organism='Human',
                          outdir=None)

gp.plot.bar(enrich_res.res2d, title='Pathway Enrichment')
plt.show()
```

### Conclusion
The analysis highlights key genes and pathways that respond to cadmium exposure, providing insights into the molecular mechanisms underlying olfactory impairment and recovery.
