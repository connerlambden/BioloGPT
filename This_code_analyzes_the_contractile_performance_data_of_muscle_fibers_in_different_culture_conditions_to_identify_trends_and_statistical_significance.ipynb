### Data Analysis of Muscle Fiber Contractility
This notebook analyzes the contractile performance of muscle fibers cultured in 2D and 3D environments.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Days': ['Day 1', 'Day 7'], '2D': [100, 50], '3D': [100, 75]}
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Days'], df['2D'], label='2D Culture')
plt.bar(df['Days'], df['3D'], label='3D Culture', bottom=df['2D'])
plt.title('Contractile Function Over Time')
plt.ylabel('Contractile Function (%)')
plt.legend()
plt.show()
```
