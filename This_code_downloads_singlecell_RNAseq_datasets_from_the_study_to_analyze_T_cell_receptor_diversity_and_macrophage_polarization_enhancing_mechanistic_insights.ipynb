Begin by importing necessary libraries and downloading the single-cell RNA-seq datasets related to GBM immune profiles.

```python
import scanpy as sc
adata = sc.read_h5ad('path_to_GBM_scRNAseq_data.h5ad')
print(adata.obs[['cell_type', 'tcr_clonality']].head())
```

Proceed with clustering and visualization of cell types, highlighting T cell subsets and macrophage polarization.

```python
sc.pp.neighbors(adata)
sc.tl.umap(adata)
sc.pl.umap(adata, color=['cell_type', 'tcr_clonality'], save='umap_analysis.png')
```

Analyze differentially expressed genes to correlate with Wnt6 and PD-L1 expression across clusters.

```python
sc.tl.rank_genes_groups(adata, 'cell_type', method='t-test')
sc.pl.rank_genes_groups(adata, n_genes=10, sharey=False, save='rank_genes.png')
```

Interpret the results to understand how clofazimine alters the immune landscape in GBM.

```python
# Code snippet to export key findings
df_results = sc.get.rank_genes_groups_df(adata, group='CD8+ T cells')
df_results.to_csv('cd8_tcell_differential_expression.csv')
```
