### Ligand Binding Affinity Prediction
This notebook utilizes datasets from recent studies to train models predicting ligand binding affinities.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

# Load dataset
data = pd.read_csv('ligand_binding_data.csv')
X = data.drop('binding_affinity', axis=1)
y = data['binding_affinity']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train model
model = RandomForestRegressor()
model.fit(X_train, y_train)

# Predict
predictions = model.predict(X_test)  # Predictions on test set
```

### Results
The model's performance can be evaluated using metrics such as R-squared and Mean Absolute Error.

```python
from sklearn.metrics import mean_absolute_error, r2_score

mae = mean_absolute_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
print(f'MAE: {mae}, R^2: {r2}')
```
