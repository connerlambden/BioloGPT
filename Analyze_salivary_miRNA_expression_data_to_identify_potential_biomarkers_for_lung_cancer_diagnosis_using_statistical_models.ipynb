This notebook will analyze salivary miRNA expression data from lung cancer patients and healthy controls.

```python
import pandas as pd
import statsmodels.api as sm

# Load the dataset
data = pd.read_csv('salivary_miRNA_data.csv')

# Fit a logistic regression model
X = data[['miR_20a', 'miR_221', 'let_7a_2', 'smoking_habit']]
y = data['lung_cancer']
X = sm.add_constant(X)
model = sm.Logit(y, X)
result = model.fit()
result.summary()
```

The logistic regression model will provide insights into the significance of each miRNA in predicting lung cancer.

```python
# Visualize the results
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x=result.params.index, y=result.params.values)
plt.title('Logistic Regression Coefficients for Salivary miRNAs')
plt.ylabel('Coefficient Value')
plt.show()
```
