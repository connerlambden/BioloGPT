### RNA-seq Data Analysis for Immunoglobulin Gene Expression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load RNA-seq data
rna_seq_data = pd.read_csv('babesia_rna_seq.csv')

# Filter for immunoglobulin genes
immunoglobulin_genes = rna_seq_data[rna_seq_data['gene_type'] == 'immunoglobulin']

# Plot expression levels
plt.figure(figsize=(10,6))
plt.bar(immunoglobulin_genes['gene_name'], immunoglobulin_genes['expression_level'])
plt.title('Expression Levels of Immunoglobulin Genes')
plt.xlabel('Gene Name')
plt.ylabel('Expression Level')
plt.xticks(rotation=45)
plt.show()
```

### This analysis provides insights into the expression patterns of immunoglobulin genes in response to Babesia bigemina infection.
