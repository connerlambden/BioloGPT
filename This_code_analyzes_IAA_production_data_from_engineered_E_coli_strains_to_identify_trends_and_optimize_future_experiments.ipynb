### IAA Production Analysis
This notebook analyzes the IAA production data from various engineered E. coli strains.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Strain': ['EcN-IAA1', 'EcN-IAA2', 'EcN-IAA3', 'EcN-IAA4', 'EcN-IAA6', 'EcN-IAA8', 'EcN-IAA9'],
        'IAA Production (mg/L)': [4.14, 10.65, 31.83, 20.6, 180.96, 30.95, 11.07]}

df = pd.DataFrame(data)

plt.bar(df['Strain'], df['IAA Production (mg/L)'])
plt.title('IAA Production by Engineered E. coli Strains')
plt.xlabel('Strain')
plt.ylabel('IAA Production (mg/L)')
plt.xticks(rotation=45)
plt.show()
```
