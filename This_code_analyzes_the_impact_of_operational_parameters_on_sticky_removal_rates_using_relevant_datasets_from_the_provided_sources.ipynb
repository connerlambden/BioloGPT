### Data Analysis of Sticky Removal Rates
This notebook analyzes the impact of temperature, disc clearance, and pulp consistency on the removal rates of stickies.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for analysis
data = {
    'Temperature (°C)': [70, 80, 90, 100],
    'Removal (%)': [20.1, 21.3, 24.8, 25.7]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['Temperature (°C)'], df['Removal (%)'], color='blue')
plt.title('Effect of Temperature on Sticky Removal')
plt.xlabel('Temperature (°C)')
plt.ylabel('Removal (%)')
plt.show()
```

### Discussion
This analysis provides insights into how temperature affects the removal of stickies, which is crucial for optimizing recycling processes.
