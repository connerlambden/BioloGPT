### Microbial Diversity Analysis During Larval Dispersal

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load microbial diversity data
# Example data structure: {'Sample Type': ['Adult', 'Larvae'], 'OTUs': [250, 150]}
data = pd.DataFrame({'Sample Type': ['Adult sponges', 'Brooding larvae', 'Free-swimming larvae'], 'OTUs': [221, 178, 403]})

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(data['Sample Type'], data['OTUs'], color='skyblue')
plt.title('Microbial OTUs in Different Life Stages')
plt.xlabel('Sample Type')
plt.ylabel('Number of OTUs')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Discussion
This analysis provides insights into how microbial diversity changes across different life stages of marine organisms, highlighting the impact of phagocytosis.
