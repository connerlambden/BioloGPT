### Consumer Trust Analysis
This notebook analyzes survey data on consumer trust in AI tools and human influencers.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Influencer Type': ['AI', 'Human Influencer'], 'Trust Level': [5.15, 2.56]}
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Influencer Type'], df['Trust Level'], color=['blue', 'orange'])
plt.title('Trust Levels in AI vs Human Influencers')
plt.ylabel('Trust Level (1-6)')
plt.show()
```
