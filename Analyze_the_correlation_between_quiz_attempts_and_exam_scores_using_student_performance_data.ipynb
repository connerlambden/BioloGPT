### Analyzing Student Performance Data
This analysis will explore the relationship between the number of quiz attempts and exam scores.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
quiz_attempts = [10, 15, 20, 25, 30]
exam_scores = [85, 88, 90, 92, 95]

# Create DataFrame
data = pd.DataFrame({'Quiz Attempts': quiz_attempts, 'Exam Scores': exam_scores})

# Plotting
plt.scatter(data['Quiz Attempts'], data['Exam Scores'])
plt.title('Quiz Attempts vs Exam Scores')
plt.xlabel('Number of Quiz Attempts')
plt.ylabel('Exam Scores')
plt.show()
```
