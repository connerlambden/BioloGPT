### RNA-seq Analysis of NF-kB Alternative Splicing
This notebook will guide you through the analysis of RNA-seq data to identify alternative splicing events in NF-kB genes.

```python
import pandas as pd
import numpy as np

# Load RNA-seq data
data = pd.read_csv('path_to_rna_seq_data.csv')

# Analyze splicing events
splicing_events = data[data['gene'].str.contains('NF-kB')]
splicing_events_summary = splicing_events.groupby('event_type').size()
splicing_events_summary
```

### Results
The analysis will provide a summary of alternative splicing events associated with NF-kB genes.

```python
# Visualize splicing events
import matplotlib.pyplot as plt
splicing_events_summary.plot(kind='bar')
plt.title('Alternative Splicing Events in NF-kB Genes')
plt.xlabel('Event Type')
plt.ylabel('Count')
plt.show()
```
