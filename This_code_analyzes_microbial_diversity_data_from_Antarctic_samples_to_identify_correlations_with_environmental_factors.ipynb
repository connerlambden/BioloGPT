### Microbial Diversity Analysis in Antarctica
This notebook analyzes microbial diversity data collected from the East Antarctic Plateau.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the microbial diversity dataset
data = pd.DataFrame({
    'Sample ID': ['N', 'M', 'S'],
    'Cell Density (cells/m3)': [171.62, 66.65, 0],
    'Microbial Taxa': [['Actinobacteriota', 'Firmicutes', 'Cyanobacteria', 'Proteobacteria'],
                      ['Actinobacteriota', 'Firmicutes', 'Cyanobacteria', 'Proteobacteria'],
                      ['Actinobacteriota', 'Firmicutes', 'Cyanobacteria', 'Proteobacteria']]
})

# Plotting the cell density
plt.figure(figsize=(10, 6))
sns.barplot(x='Sample ID', y='Cell Density (cells/m3)', data=data)
plt.title('Cell Density of Microbial Samples from Antarctic Plateau')
plt.ylabel('Cell Density (cells/m3)')
plt.xlabel('Sample ID')
plt.show()
```
