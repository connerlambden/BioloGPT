### Analyzing the Impact of Digital Platforms on Tax Revenues
This analysis will utilize data from the study on digital platforms in tax administration services.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data based on the provided research
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    '2-3 Wheels': [6.4, 7.8, 8.2, 9.3, 10.4],
    '≥ 4 Wheels': [1.4, 1.7, 2.3, 2.8, 3.2],
    'Total Users': [7.8, 8.15, 10.5, 11.11, 13.6]
}

df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['2-3 Wheels'], marker='o', label='2-3 Wheels')
plt.plot(df['Year'], df['≥ 4 Wheels'], marker='o', label='≥ 4 Wheels')
plt.plot(df['Year'], df['Total Users'], marker='o', label='Total Users')
plt.title('Impact of Digital Platforms on Tax Administration Services')
plt.xlabel('Year')
plt.ylabel('Number of Users (in millions)')
plt.legend()
plt.grid()
plt.show()
```
