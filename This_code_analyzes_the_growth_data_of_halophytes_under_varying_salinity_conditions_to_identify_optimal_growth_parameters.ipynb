### Halophyte Growth Analysis
This notebook analyzes the growth data of halophytes under different salinity levels.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for halophyte growth
data = {'Species': ['Atriplex lentiformis', 'Atriplex nummularia', 'Atriplex halimus', 'Sporobolus'],
        'Productivity (t/ha)': [25.0, 16.9, 14.6, 29]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['Species'], df['Productivity (t/ha)'], color='green')
plt.title('Productivity of Selected Halophytes')
plt.xlabel('Species')
plt.ylabel('Productivity (t/ha)')
plt.show()
```
