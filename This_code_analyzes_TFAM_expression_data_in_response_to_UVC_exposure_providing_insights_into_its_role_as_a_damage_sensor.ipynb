### TFAM Expression Analysis
This notebook analyzes the expression levels of TFAM in response to varying doses of UVC irradiation.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for TFAM expression
data = {'UVC_dose': ['0 J/m2', '10 J/m2', '30 J/m2'], 'TFAM_expression': [1, 2, 3]}
df = pd.DataFrame(data)

# Plotting the data
plt.bar(df['UVC_dose'], df['TFAM_expression'], color='blue')
plt.title('TFAM Expression in Response to UVC Doses')
plt.xlabel('UVC Dose')
plt.ylabel('TFAM Expression Level')
plt.show()
```

This analysis highlights the correlation between UVC exposure and TFAM expression levels.
