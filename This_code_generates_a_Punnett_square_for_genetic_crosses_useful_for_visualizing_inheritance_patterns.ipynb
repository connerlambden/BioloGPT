### Punnett Square Generator for Guinea Pig Genetics
This notebook will create a Punnett square for the genetic cross of guinea pigs with specified genotypes.

```python
import pandas as pd
import numpy as np

# Define the genotypes
parent1 = ['N', 'N']  # White guinea pig (NN)
parent2 = ['N', 'n']  # Black guinea pig (Nn)

# Create the Punnett square
punnett_square = pd.DataFrame(index=parent1, columns=parent2)

for i in parent1:
    for j in parent2:
        punnett_square.loc[i, j] = i + j

punnett_square.fillna('', inplace=True)
punnett_square
```

### Resulting Punnett Square
The resulting Punnett square will show the possible genotypes of the offspring.

```python
punnett_square
```
