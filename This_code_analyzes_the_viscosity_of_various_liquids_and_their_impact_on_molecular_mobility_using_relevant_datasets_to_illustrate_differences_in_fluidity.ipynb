### Analyzing Liquid Viscosity and Fluidity
This notebook will explore the relationship between viscosity and fluidity in various liquids.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for viscosity of different liquids
data = {'Liquid': ['Water', 'Oil', 'Glycerin'], 'Viscosity (mPa.s)': [0.89, 50, 1200]}
df = pd.DataFrame(data)

# Plotting the viscosity
plt.bar(df['Liquid'], df['Viscosity (mPa.s)'], color=['blue', 'yellow', 'green'])
plt.title('Viscosity of Different Liquids')
plt.xlabel('Liquid')
plt.ylabel('Viscosity (mPa.s)')
plt.show()
```

### Discussion
This analysis shows how different liquids exhibit varying viscosities, impacting their fluidity and behavior in biological systems.
