### Analysis of Multi-resolution ViT Model Performance
This notebook analyzes the classification performance of the multi-resolution vision transformer model across different magnifications.

```python
import pandas as pd
import plotly.express as px

def load_performance_data():
    # Assuming the data is in a CSV file
    url = 'https://biologpt.com/data/performance_metrics.csv'
    df = pd.read_csv(url)
    return df
```

### Load and Inspect Data

```python
df = load_performance_data()
df.head()
```

### Visualize Kappa Scores Across Resolutions

```python
fig = px.bar(df, x='Resolution', y='Kappa_Score', color='Subtype', barmode='group', title='Kappa Scores by Resolution and Subtype')
fig.show()
```

### Confusion Matrix Visualization for Melanoma Classification

```python
import plotly.figure_factory as ff

# Example confusion matrix data
conf_matrix = [[945, 55], [30, 970]]
fig = ff.create_annotated_heatmap(z=conf_matrix, x=['Predicted Melanoma', 'Predicted Non-Melanoma'], y=['Actual Melanoma', 'Actual Non-Melanoma'], colorscale='Blues')
fig.update_layout(title_text='Confusion Matrix for Melanoma Classification')
fig.show()
```
