This notebook downloads MD trajectory data and uses analysis packages to assess secondary structure evolution and lipid interaction frequencies.

```python
import MDAnalysis as mda
import numpy as np
import matplotlib.pyplot as plt

# Load simulation trajectory and topology
u = mda.Universe('topology.pdb', 'trajectory.dcd')

# Define analysis: calculate contacts between peptide and lipid headgroups
contacts = []
for ts in u.trajectory:
    distance = np.linalg.norm(u.select_atoms('resname POPE or resname DOPG').positions - u.select_atoms('protein').positions.mean(axis=0))
    contacts.append(distance)

plt.plot(contacts)
plt.xlabel('Frame')
plt.ylabel('Peptide-Lipid Contact Distance')
plt.title('MD Simulation Analysis of Peptide-Lipid Interactions')
plt.show()
```

This plot assesses distance dynamics as a proxy for interaction strength between the peptide and the lipid bilayer.
