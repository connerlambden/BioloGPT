This cell loads the dataset containing athlete measures including DXA LST, D3Cr SMM, and muscle strength values, then prepares the data for visualization.

```python
import pandas as pd
import plotly.express as px

# Load dataset sample
data = pd.DataFrame({
    'Measure': ['Leg Strength', 'Trunk Strength', 'DXA LST', 'D3Cr SMM'],
    'Mean': [138.36, 184.77, 70.43, 41.3]
})

fig = px.bar(data, x='Measure', y='Mean', title='Mean Muscle Strength and Body Composition Values',
             color='Measure', color_discrete_sequence=['#6A0C76', '#6A0C76', '#f39c12', '#f39c12'])
fig.show()
```
