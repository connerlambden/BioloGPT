### Simulating Genetic Gains with Additional Phenotyping
This notebook will analyze the impact of additional phenotyping on genetic gains in maize breeding using MoBPS data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for genetic gains
data = {'Cycle': [1, 30], 'Genetic_Gain_Cross': [6.8, 9.34], 'Genetic_Gain_Per_Se': [6.95, 7.89]}
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(df['Cycle'], df['Genetic_Gain_Cross'], marker='o', label='Genetic Gain (Cross)', color='blue')
plt.plot(df['Cycle'], df['Genetic_Gain_Per_Se'], marker='o', label='Genetic Gain (Per Se)', color='green')
plt.title('Genetic Gains Over Breeding Cycles')
plt.xlabel('Breeding Cycle')
plt.ylabel('Genetic Gain (gSD)')
plt.legend()
plt.grid()
plt.show()
```
