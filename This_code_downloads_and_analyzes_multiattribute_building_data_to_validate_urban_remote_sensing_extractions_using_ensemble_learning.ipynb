Step-by-step instructions for downloading the CMAB dataset and performing data quality validation.

```python
import requests
import pandas as pd
# Download dataset from Figshare
url = 'https://doi.org/10.6084/m9.figshare.27992417.v2'
df = pd.read_csv(url)
# Perform basic validations
df.head()
print('Dataset shape:', df.shape)
```

Visualize the distribution of building heights and functions using Plotly.

```python
import plotly.express as px
fig = px.histogram(df, x='height', color='function', barmode='overlay', title='Distribution of Building Heights by Function')
fig.show()
```
