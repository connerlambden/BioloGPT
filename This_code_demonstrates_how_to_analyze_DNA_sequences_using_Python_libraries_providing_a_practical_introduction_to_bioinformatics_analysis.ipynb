### Step 1: Import Required Libraries
This step involves importing necessary libraries for bioinformatics analysis.

```python
import pandas as pd
from Bio import SeqIO
import matplotlib.pyplot as plt
```

### Step 2: Load DNA Sequence Data
Load your DNA sequence data from a FASTA file.

```python
sequences = list(SeqIO.parse('your_sequences.fasta', 'fasta'))
```

### Step 3: Analyze Sequence Lengths
Calculate the lengths of the sequences and visualize the distribution.

```python
lengths = [len(seq.seq) for seq in sequences]
plt.hist(lengths, bins=20)
plt.title('Distribution of DNA Sequence Lengths')
plt.xlabel('Length')
plt.ylabel('Frequency')
plt.show()
```

### Step 4: Summary Statistics
Generate summary statistics for the sequence lengths.

```python
print(f'Mean Length: {pd.Series(lengths).mean()}')
print(f'Median Length: {pd.Series(lengths).median()}')
print(f'Max Length: {pd.Series(lengths).max()}')
```
