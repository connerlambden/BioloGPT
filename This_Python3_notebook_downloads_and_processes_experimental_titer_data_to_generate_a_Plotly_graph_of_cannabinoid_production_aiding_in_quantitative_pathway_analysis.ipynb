This section describes the setup to analyze the experimental production titers reported in the paper. The dataset includes cannabinoids and their measured titers. We use Plotly to create an interactive graph for visual inspection.

```python
import plotly.graph_objs as go
import plotly.offline as pyo

# Define data
cannabinoids = ['CBGA', 'Orsellinic Acid', 'Cannabigerorcinic Acid']
titers = [15.7, 18.8, 0.5]

# Create bar chart
data = [go.Bar(x=cannabinoids, y=titers, marker=dict(color='#6A0C76'))]

layout = go.Layout(
    title='Cannabinoid Production Titers in Yarrowia lipolytica',
    xaxis=dict(title='Cannabinoid'),
    yaxis=dict(title='Titer (mg/L)')
)

fig = go.Figure(data=data, layout=layout)
pyo.plot(fig, filename='cannabinoid_titers.html')
```

The above code downloads the relevant experimental titer data and generates an interactive bar graph using Plotly. This visualization helps in comparing the production levels of different cannabinoids.
