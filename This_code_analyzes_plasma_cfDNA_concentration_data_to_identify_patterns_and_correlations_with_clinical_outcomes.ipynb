### Data Analysis of Plasma cfDNA Concentrations

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Cohort': ['Acute Aortic', 'Chronic Aortic', 'Control'], 'cfDNA Concentration': [75, 30, 20]}
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Cohort'], df['cfDNA Concentration'], color=['red', 'blue', 'green'])
plt.title('Plasma cfDNA Concentration by Cohort')
plt.xlabel('Cohort')
plt.ylabel('cfDNA Concentration (ng/uL)')
plt.show()
```

This analysis provides insights into the potential of cfDNA as a biomarker for aortic diseases.
