### Gene Expression Analysis of T and B Cells

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Gene': ['CD4', 'HLA-DOB', 'HLA-DRB1', 'HLA-DRA', 'AHNAK'], 'Expression': [-2.7, -2.2, -2.8, -2.2, -2.0]}
df = pd.DataFrame(data)
plt.bar(df['Gene'], df['Expression'])
plt.title('Gene Expression Levels in T and B Cells')
plt.xlabel('Genes')
plt.ylabel('Log2 Fold Change')
plt.show()
```

This analysis provides insights into the expression levels of key immune genes in T and B cells.
