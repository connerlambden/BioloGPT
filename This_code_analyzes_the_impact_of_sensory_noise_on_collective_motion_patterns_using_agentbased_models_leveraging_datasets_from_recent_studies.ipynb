### Analysis of Sensory Noise Impact on Collective Motion Patterns
This notebook will utilize datasets from recent studies to model the effects of sensory noise on collective motion.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate collective motion under varying noise levels
noise_levels = np.linspace(0, 1, 10)
coherence = np.exp(-noise_levels)  # Hypothetical coherence function

plt.figure(figsize=(10, 6))
plt.plot(noise_levels, coherence, marker='o')
plt.title('Impact of Sensory Noise on Collective Motion Coherence')
plt.xlabel('Sensory Noise Level')
plt.ylabel('Coherence')
plt.grid(True)
plt.show()
```
