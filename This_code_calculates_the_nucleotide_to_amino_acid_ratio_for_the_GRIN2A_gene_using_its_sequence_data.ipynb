### Step 1: Import Required Libraries
Import necessary libraries for handling biological sequences.

```python
from Bio import SeqIO

# Load the GRIN2A gene sequence
sequence = SeqIO.read('grin2a_sequence.fasta', 'fasta')
```

### Step 2: Calculate the Ratio
Calculate the number of nucleotides and amino acids.

```python
# Calculate the number of nucleotides
num_nucleotides = len(sequence.seq)

# Calculate the number of amino acids (assuming full coding sequence)
num_amino_acids = num_nucleotides // 3

# Calculate the ratio
ratio = num_nucleotides / num_amino_acids
ratio
```

### Step 3: Output the Result
Display the calculated ratio.

```python
print(f'The ratio of nucleotides to amino acids in GRIN2A is: {ratio}:1')
```
