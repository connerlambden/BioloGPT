Import and preprocess the patient dataset to generate a scatter plot correlating age and microbial isolate findings.

```python
import pandas as pd
import plotly.express as px

data = [{'patient_id': 1, 'age': 39, 'microorganism': 'MRS'},
        {'patient_id': 2, 'age': 27, 'microorganism': 'MRS'},
        {'patient_id': 3, 'age': 33, 'microorganism': 'Micrococcus'},
        {'patient_id': 4, 'age': 41, 'microorganism': 'Diphtheroid'},
        {'patient_id': 5, 'age': 36, 'microorganism': 'CoNS'},
        {'patient_id': 6, 'age': 44, 'microorganism': 'CoNS'},
        {'patient_id': 7, 'age': 25, 'microorganism': 'E. coli'}]

ndf = pd.DataFrame(data)
fig = px.scatter(ndf, x='age', y='patient_id', color='microorganism', title='Patient Age vs Microbial Isolates')
fig.show()
```
