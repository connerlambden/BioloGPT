### Analyzing the Impact of pCO2 on Foraminiferal Morphology
This notebook will utilize the data extracted from the study to analyze the effects of varying pCO2 levels on the morphology of Haynesina sp.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from the study
data = {
    'Trial': ['Summer 2023', 'Fall 2023'],
    'pCO2 (μatm)': [4495.21, 5100.08],
    'Ωcalcite': [0.865, 0.843]
}

df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 5))
plt.bar(df['Trial'], df['pCO2 (μatm)'], color='blue', alpha=0.7, label='pCO2 (μatm)')
plt.ylabel('pCO2 (μatm)')
plt.title('pCO2 Levels in Different Trials')
plt.legend()
plt.show()
```
