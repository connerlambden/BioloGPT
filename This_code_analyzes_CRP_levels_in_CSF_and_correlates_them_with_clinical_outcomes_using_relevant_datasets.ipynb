### CRP Level Analysis in CSF
This notebook analyzes CRP levels in cerebrospinal fluid (CSF) and their correlation with clinical outcomes in neurological diseases.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for CRP levels
data = {'Group': ['Control', 'SAH Day 1', 'SAH Day 4', 'SAH Day 7', 'SAH Day 10'],
        'CRP Level (mg/L)': [1.15, 5.99, 13.83, 15.17, 25.82]}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Group'], df['CRP Level (mg/L)'])
plt.title('CRP Levels in CSF Over Time')
plt.xlabel('Patient Group')
plt.ylabel('CRP Level (mg/L)')
plt.show()
```

### Discussion
This analysis provides insights into the inflammatory response in neurological conditions and the role of CRP as a biomarker.
