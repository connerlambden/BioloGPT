This notebook will analyze menstrual cycle data to explore correlations between cycle phases and reported symptoms.

```python
import pandas as pd

# Load menstrual cycle dataset
dataset = pd.read_csv('menstrual_cycle_data.csv')

# Analyze correlations
correlations = dataset.corr()
```

The analysis will provide insights into how different phases of the menstrual cycle correlate with various health outcomes.

```python
# Display correlations
correlations.to_csv('menstrual_cycle_correlations.csv')
```
