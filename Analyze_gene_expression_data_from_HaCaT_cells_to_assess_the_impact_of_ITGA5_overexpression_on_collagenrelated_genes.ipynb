This notebook will analyze RNA-seq data from HaCaT cells overexpressing ITGA5 to identify changes in collagen gene expression.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('HaCaT_ITGA5_expression.csv')

# Visualize the expression levels of collagen genes
collagen_genes = ['COL1A1', 'COL3A1', 'COL5A1']
expression_data = data[collagen_genes]
sns.boxplot(data=expression_data)
plt.title('Collagen Gene Expression in HaCaT Cells')
plt.show()
```

This analysis will help determine the effect of ITGA5 overexpression on collagen synthesis in HaCaT cells.
