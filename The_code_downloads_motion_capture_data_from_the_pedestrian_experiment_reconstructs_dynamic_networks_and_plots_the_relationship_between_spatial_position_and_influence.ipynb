This notebook downloads the real motion capture dataset and processes the data to reconstruct the visual influence network using TDDC.

```python
import pandas as pd
import numpy as np
# Code to download and load the dataset
url = 'https://doi.org/10.26300/xm9v-tg76'
data = pd.read_csv(url)
print(data.head())
```

Next, we compute network measures like direct and branching influence, and plot the influence gradient as a function of spatial position.

```python
import plotly.express as px
# Assuming data has columns 'position' and 'influence'
fig = px.scatter(data, x='position', y='influence', trendline='ols', title='Influence vs Spatial Position')
fig.show()
```
