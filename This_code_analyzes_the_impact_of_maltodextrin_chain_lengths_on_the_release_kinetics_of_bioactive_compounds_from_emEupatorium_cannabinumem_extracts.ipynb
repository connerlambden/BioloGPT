Analyze the release profiles of bioactive compounds from maltodextrin-encapsulated <em>Eupatorium cannabinum</em> extracts.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data on release profiles
release_data = pd.read_csv('release_profiles.csv')

# Plot release profiles
plt.figure(figsize=(10, 6))
for chain_length in release_data['Chain Length'].unique():
    subset = release_data[release_data['Chain Length'] == chain_length]
    plt.plot(subset['Time'], subset['Release'], label=f'Chain Length: {chain_length}')
plt.title('Release Profiles of Bioactive Compounds')
plt.xlabel('Time (hours)')
plt.ylabel('Release (%)')
plt.legend()
plt.show()
```

This analysis will help visualize how different maltodextrin chain lengths affect the release of bioactive compounds.
