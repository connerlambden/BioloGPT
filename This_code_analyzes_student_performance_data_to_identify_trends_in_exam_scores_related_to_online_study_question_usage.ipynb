### Analyzing Exam Performance Data
This notebook will analyze the exam performance data of students who used online study questions versus those who did not.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
exam_data = pd.read_csv('exam_scores.csv')

# Analyze the data
performance_summary = exam_data.groupby('Year').mean()
performance_summary.plot(kind='bar')
plt.title('Average Exam Scores Over Years')
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.show()
```

### Summary of Findings
The analysis will provide insights into how online study questions impact overall student performance.

```python
# Save the performance summary for further analysis
performance_summary.to_csv('performance_summary.csv')
```
