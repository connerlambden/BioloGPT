This notebook simulates blood type inheritance based on parental genotypes.

```python
# Simulate blood type inheritance
import random

# Define parent genotypes
parent1 = ['IA', 'i']  # Peeta's genotype
parent2 = ['IB', 'IB']  # Katniss's genotype

# Simulate offspring genotype
offspring_genotype = random.choice(parent1) + random.choice(parent2)
print('Offspring genotype:', offspring_genotype)  # Display offspring genotype
```

This code randomly selects alleles from each parent to simulate the potential genotype of their offspring.

```python
# Display possible blood types based on offspring genotype
blood_types = {'IAIB': 'AB', 'IBi': 'B'}
print('Possible blood type:', blood_types.get(offspring_genotype, 'Unknown'))
```
