This notebook loads experimental data on PHBV production and visualizes the effect of nitrogen source and C/N ratio on polymer accumulation and HV content.

```python
import pandas as pd
import plotly.express as px

# Experimental data based on extracted study details
data = [
    {'Experiment': 'A', 'Nitrogen_Source': 'Ammonium', 'C/N_Ratio': 8, 'Biomass_g_per_L': 11.3, 'PHA_g_per_L': 0.63, 'PHA_Accumulation_%': 4.6, 'HV_%': 16.9},
    {'Experiment': 'A', 'Nitrogen_Source': 'Nitrate', 'C/N_Ratio': 8, 'Biomass_g_per_L': 6.4, 'PHA_g_per_L': 0.80, 'PHA_Accumulation_%': 9.3, 'HV_%': 12.5},
    {'Experiment': 'B', 'Nitrogen_Source': None, 'C/N_Ratio': 25, 'PHA_Accumulation_%': 4.8},
    {'Experiment': 'B', 'Nitrogen_Source': None, 'C/N_Ratio': 42, 'PHA_Accumulation_%': 6.6}
]
df = pd.DataFrame(data)

fig = px.bar(df, x='Experiment', y='PHA_Accumulation_%', color='Nitrogen_Source', barmode='group',
             title='PHBV Accumulation by Nitrogen Source and C/N Ratio')
fig.show()
```

The above graph visualizes how different nitrogen sources and C/N ratios affect PHBV accumulation, an essential step towards optimizing bioprocess conditions.

```python
import statsmodels.api as sm

# Remove rows with missing data for regression analysis
df_clean = df.dropna(subset=['C/N_Ratio', 'PHA_Accumulation_%'])
X = df_clean[['C/N_Ratio']]
Y = df_clean['PHA_Accumulation_%']
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
print(model.summary())
```
