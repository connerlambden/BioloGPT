### Step 1: Import Necessary Libraries
Import libraries such as pandas, numpy, and scipy for data manipulation and analysis.

```python
import pandas as pd
import numpy as np
from scipy import stats
```

### Step 2: Load Genetic Data
Load the genetic dataset containing information on variants and their frequencies.

```python
genetic_data = pd.read_csv('genetic_variants.csv')
```

### Step 3: Identify Rare Variants
Filter the dataset to identify rare variants based on a defined frequency threshold.

```python
rare_variants = genetic_data[genetic_data['frequency'] < 0.01]
```

### Step 4: Statistical Analysis
Perform statistical tests to assess the association of rare variants with complex diseases.

```python
results = stats.ttest_ind(rare_variants['disease_positive'], rare_variants['disease_negative'])
```

### Step 5: Output Results
Display the results of the statistical analysis.

```python
print('T-statistic:', results.statistic)
print('P-value:', results.pvalue)
```
