Load AML dataset (GEO GSE288783) and perform differential expression analysis of eIF4E-regulated mRNAs.

```python
import pandas as pd
import plotly.express as px
# Load dataset
df = pd.read_csv('path/to/GSE288783_expression.csv')
# Create a scatter plot of log2 fold changes vs p-values
fig = px.scatter(df, x='log2FoldChange', y='-log10(pvalue)', color='gene', title='Differential Expression of eIF4E Target Genes')
fig.show()
```
