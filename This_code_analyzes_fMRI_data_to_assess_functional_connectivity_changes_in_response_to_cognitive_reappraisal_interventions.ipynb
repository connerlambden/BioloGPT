### Data Preparation
Load fMRI data and preprocess for analysis.

```python
import nibabel as nib
import numpy as np

# Load fMRI data
fmri_data = nib.load('path_to_fmri_data.nii').get_fdata()
```

### Functional Connectivity Analysis
Calculate functional connectivity matrices.

```python
from sklearn.covariance import GraphicalLasso

# Calculate connectivity
connectivity_matrix = GraphicalLasso().fit(fmri_data.reshape(-1, fmri_data.shape[-1])).covariance_
```

### Visualization
Visualize the connectivity matrix.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(connectivity_matrix, cmap='coolwarm')
plt.title('Functional Connectivity Matrix')
plt.show()
```
