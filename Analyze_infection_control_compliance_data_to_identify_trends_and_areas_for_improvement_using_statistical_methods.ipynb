This notebook will analyze compliance data from various ICUs to identify factors influencing adherence to infection control practices.

```python
import pandas as pd

# Load compliance data
data = pd.read_csv('compliance_data.csv')

# Analyze trends in adherence
compliance_trends = data.groupby('ICU').mean()
compliance_trends.to_csv('compliance_trends.csv')
```

The analysis will provide insights into which ICUs are performing well and which need additional support.

```python
# Visualize compliance trends
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(compliance_trends.index, compliance_trends['adherence'])
plt.title('Compliance Trends by ICU')
plt.xlabel('ICU')
plt.ylabel('Average Adherence')
plt.show()
```
