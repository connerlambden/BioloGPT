This notebook will analyze climate data (temperature, precipitation) alongside allergen data (pollen counts) to identify correlations.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load climate and allergen data
data = pd.read_csv('vaal_triangle_data.csv')

# Analyze correlations
correlation = data.corr()

# Visualize results
plt.figure(figsize=(10, 6))
plt.title('Correlation between Climate Variables and Allergen Levels')
plt.imshow(correlation, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.show()
```

This analysis will help identify key climate factors influencing allergen levels.
