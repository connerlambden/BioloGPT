### Analyzing Active Learning Impact on Exam Performance
This analysis will utilize student performance data to evaluate the effectiveness of online study questions.

```python
import pandas as pd

# Load student performance data
data = pd.read_csv('student_performance.csv')

# Calculate average scores for students with and without access to study questions
average_scores = data.groupby('access_to_study_questions')['exam_score'].mean()
print(average_scores)
```
