This notebook will analyze metagenomic sequences from urban soil samples to identify microbial taxa associated with bioremediation.

```python
import pandas as pd

# Load urban soil microbiome data
data = pd.read_csv('urban_soil_microbiome.csv')

# Analyze key taxa involved in pollutant degradation
key_taxa = data[data['function'] == 'pollutant_degradation']
key_taxa_summary = key_taxa.groupby('taxa').size().reset_index(name='counts')
key_taxa_summary.sort_values(by='counts', ascending=False)
```
