This analysis will utilize datasets on receptor localization to explore the distribution of membrane-bound and intracellular receptors.

```python
import pandas as pd

# Load receptor localization data
data = pd.read_csv('receptor_localization.csv')

# Analyze the distribution of receptor types
membrane_bound = data[data['location'] == 'membrane']
intracellular = data[data['location'] == 'intracellular']

# Summary statistics
summary = {
    'membrane_bound_count': len(membrane_bound),
    'intracellular_count': len(intracellular)
}

summary
```

```python
# Display summary statistics
print(summary)
```
