### Molar Mass Calculation of CuCO₃ • Cu(OH)₂
This notebook calculates the molar mass of the compound using atomic weights.

```python
# Define atomic weights
atomic_weights = {'Cu': 63.55, 'C': 12.01, 'O': 16.00, 'H': 1.01}

# Calculate molar mass of CuCO3
CuCO3_mass = atomic_weights['Cu'] + atomic_weights['C'] + 3 * atomic_weights['O']

# Calculate molar mass of Cu(OH)2
CuOH2_mass = atomic_weights['Cu'] + 2 * atomic_weights['O'] + 2 * atomic_weights['H']

# Total mass
total_mass = CuCO3_mass + CuOH2_mass

total_mass # Output the total mass
```

### Result
The total molar mass of CuCO₃ • Cu(OH)₂ is calculated.

```python
# Display the result
print(f'Total Molar Mass of CuCO₃ • Cu(OH)₂: {total_mass:.2f} g/mol')
```
