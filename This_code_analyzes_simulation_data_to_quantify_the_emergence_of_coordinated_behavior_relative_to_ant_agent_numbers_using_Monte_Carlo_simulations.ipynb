Import simulation data and calculate the emergence function across different ant densities.

```python
import numpy as np
import matplotlib.pyplot as plt

# Example simulation data
agents = np.array([1000, 2000, 3000, 4000, 5000, 6000, 7000])
coupling_efficiency = np.array([0.2, 0.35, 0.55, 0.70, 0.90, 0.93, 0.95])

plt.figure(figsize=(8, 5))
plt.plot(agents, coupling_efficiency, marker='o', color='#6A0C76')
plt.title('Emergence of Coordinated Behavior vs Number of Ants')
plt.xlabel('Number of Ant Agents')
plt.ylabel('Coordination Efficiency')
plt.grid(True)
plt.show()
```

The code above provides a simple visualization of how increased ant numbers correlate with coordination efficiency, supporting the VDR model.

```python
# Calculate threshold value
threshold_index = np.where(coupling_efficiency >= 0.9)[0][0]
print('Threshold number of agents:', agents[threshold_index])
```
