### Step 1: Import Libraries
Import necessary libraries for data analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load SNP Data
Load the SNP data for argali subspecies.

```python
snp_data = pd.read_csv('argali_snp_data.csv')
snp_data.head()
```

### Step 3: Analyze Genetic Diversity
Calculate genetic diversity metrics for each subspecies.

```python
def calculate_diversity(data):
    diversity = data.apply(lambda x: np.nanmean(x), axis=0)
    return diversity

genetic_diversity = calculate_diversity(snp_data)
print(genetic_diversity)
```
