### Antimicrobial Activity Analysis
This notebook analyzes the antimicrobial activity of magnolol and honokiol formulations against various pathogens.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for analysis
data = {
    'Formulation': ['β-CDLA/MG', 'β-CDLA/HK', 'Pure MG', 'Pure HK'],
    'S. aureus (mm)': [20.0, 20.1, 20.0, 23.0],
    'E. coli (mm)': [20.1, 15.0, None, None],
    'C. albicans (mm)': [14.7, 15.0, 16.0, 18.0]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 6))
for column in df.columns[1:]:
    plt.plot(df['Formulation'], df[column], marker='o', label=column)
plt.title('Antimicrobial Activity of Formulations')
plt.xlabel('Formulation')
plt.ylabel('Inhibition Zone (mm)')
plt.legend()
plt.grid()
plt.show()
```
