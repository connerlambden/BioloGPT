# TOM500 Dataset Analysis
This notebook analyzes the TOM500 dataset, focusing on clinical characteristics and segmentation performance.

```python
import pandas as pd

dataset = pd.read_csv('TOM500_data.csv')
clinical_data = dataset[['sex', 'age', 'disease_duration', 'smoking_status']]
segmentation_performance = dataset[['Dice', 'Jaccard', 'Sensitivity', 'Specificity', 'Precision', 'Recall', 'Accuracy']]

clinical_data.describe()
segmentation_performance.describe()
```

# Summary of Clinical Characteristics and Segmentation Performance
This section summarizes the key findings from the dataset.

```python
clinical_data_summary = clinical_data.groupby('sex').count()
segmentation_performance_summary = segmentation_performance.mean()
```
