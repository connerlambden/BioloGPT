Step 1: Load and preprocess scRNA-seq data from the study's experiments.

```python
import scanpy as sc
adata = sc.read('path_to_scRNAseq_data.h5ad')
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
adata.write('preprocessed_scRNAseq.h5ad')
```

Step 2: Visualize key gene expressions, including NF-κB pathway genes.

```python
import matplotlib.pyplot as plt
sc.pl.violin(adata, keys=['NFKB1', 'NFKB2'], groupby='treatment', jitter=0.4)
plt.show()
```
