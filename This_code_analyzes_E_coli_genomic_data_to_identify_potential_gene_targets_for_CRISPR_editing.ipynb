### E. coli Genomic Data Analysis

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load genomic data
data = pd.read_csv('ecoli_genome_data.csv')

# Analyze gene expression levels
expression_levels = data.groupby('gene')['expression'].mean()
```

### Summary of Gene Expression Levels

```python
# Display summary statistics
summary = expression_levels.describe()
print(summary)
```
