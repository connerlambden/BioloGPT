### Step 1: Data Collection
Import the RdRp gene sequences from GenBank.

```python
import pandas as pd
from Bio import SeqIO
from Bio.Align.Applications import ClustalOmegaCommandline
from Bio import Phylo
import matplotlib.pyplot as plt

# Load sequences
sequences = list(SeqIO.parse('RdRp_sequences.fasta', 'fasta'))
```

### Step 2: Sequence Alignment
Use Clustal Omega for multiple sequence alignment.

```python
cline = ClustalOmegaCommandline(infile='RdRp_sequences.fasta', outfile='aligned_RdRp.aln', verbose=True, auto=True)
cline()

# Read alignment
alignment = SeqIO.read('aligned_RdRp.aln', 'clustal')
```

### Step 3: Phylogenetic Tree Construction
Construct a phylogenetic tree using the aligned sequences.

```python
from Bio.Phylo.TreeConstruction import DistanceCalculator, DistanceTreeConstructor

calculator = DistanceCalculator('identity')
dm = calculator.get_distance(alignment)
constructor = DistanceTreeConstructor()
tree = constructor.nj(dm)
Phylo.draw(tree)
plt.show()
```
