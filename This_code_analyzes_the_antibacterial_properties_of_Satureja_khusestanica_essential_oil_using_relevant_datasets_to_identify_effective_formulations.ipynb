This notebook will explore the antibacterial activity of Satureja khusestanica essential oil against various bacterial strains.

```python
import pandas as pd
# Load relevant datasets
datasets = pd.read_csv('satureja_khusestania_data.csv')
# Analyze antibacterial activity
results = datasets.groupby('bacterial_strain')['activity'].mean()
```

The results will provide insights into the effectiveness of the essential oil against different bacterial strains.

```python
results.to_csv('antibacterial_results.csv') # Save the results for further analysis.
```
