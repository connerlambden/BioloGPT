This notebook will analyze longitudinal sNfL data to identify patterns and predict cognitive outcomes in RRMS patients.

```python
import pandas as pd
import statsmodels.api as sm

# Load sNfL data
snfl_data = pd.read_csv('snfl_data.csv')

# Fit a linear regression model to predict cognitive outcomes based on sNfL levels
model = sm.OLS(snfl_data['cognitive_outcome'], sm.add_constant(snfl_data['sNfL'])).fit()
model.summary()
```

The model summary will provide insights into the relationship between sNfL levels and cognitive outcomes.

```python
# Visualize the relationship
import matplotlib.pyplot as plt
plt.scatter(snfl_data['sNfL'], snfl_data['cognitive_outcome'])
plt.xlabel('sNfL Levels')
plt.ylabel('Cognitive Outcome')
plt.title('sNfL Levels vs Cognitive Outcome')
plt.show()
```
