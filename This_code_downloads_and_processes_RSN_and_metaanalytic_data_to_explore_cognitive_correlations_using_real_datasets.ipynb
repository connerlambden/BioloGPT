Step-by-step instructions on accessing the GINNA dataset, preprocessing using ICA, and comparing cognitive labels from meta-analysis.

```python
import pandas as pd
import numpy as np
# Download RSN data from repository
rsn_data = pd.read_csv('https://github.com/Achillegillig/ginna/raw/main/rsn_data.csv')
# Process meta-analytic decoding results
decoding_results = pd.read_csv('https://github.com/Achillegillig/ginna/raw/main/decoding_results.csv')
# Merge and perform analysis
merged = pd.merge(rsn_data, decoding_results, on='RSN_ID')
print(merged.head())
```

The notebook details data acquisition, merging, and basic statistical analysis to validate cognitive labels across RSNs.

```python
import matplotlib.pyplot as plt
plt.bar(merged['RSN_Label'], merged['Correlation'])
plt.xlabel('RSN Cognitive Label')
plt.ylabel('Correlation Coefficient')
plt.title('RSN Cognitive Correlation Analysis')
plt.xticks(rotation=45)
plt.show()
```
