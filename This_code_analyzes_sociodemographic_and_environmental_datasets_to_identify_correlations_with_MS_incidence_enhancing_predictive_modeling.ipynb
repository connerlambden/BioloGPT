### Data Analysis of MS Incidence
This notebook will analyze socio-demographic and environmental data to refine predictive models for MS incidence.

```python
import pandas as pd
import statsmodels.api as sm

# Load datasets
ms_data = pd.read_csv('ms_incidence_data.csv')
socio_demo_data = pd.read_csv('socio_demo_data.csv')

# Merge datasets
merged_data = pd.merge(ms_data, socio_demo_data, on='region')

# Fit a regression model
X = merged_data[['socio_demo_var1', 'socio_demo_var2', 'environmental_var1']]
y = merged_data['ms_incidence']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

# Print the summary
model.summary()
```

### Results Interpretation
The results will provide insights into which socio-demographic and environmental factors significantly predict MS incidence.

```python
# Visualizing the results
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x=model.params.index, y=model.params.values)
plt.title('Impact of Socio-Demographic and Environmental Factors on MS Incidence')
plt.ylabel('Coefficient')
plt.xticks(rotation=45)
plt.show()
```
