### Data Preparation
Load and preprocess battery performance and thermal data.

```python
import pandas as pd

# Load datasets
df_battery = pd.read_csv('battery_data.csv')
df_thermal = pd.read_csv('thermal_data.csv')

# Merge datasets based on timestamps
df_merged = pd.merge(df_battery, df_thermal, on='timestamp')
```

### Model Training
Train the IRF algorithm with integrated thermal data.

```python
from sklearn.ensemble import RandomForestRegressor

# Define features and target
X = df_merged[['voltage', 'current', 'temperature', 'cycle_count']]
y = df_merged['state_of_health']

# Train the model
model = RandomForestRegressor()
model.fit(X, y)
```
