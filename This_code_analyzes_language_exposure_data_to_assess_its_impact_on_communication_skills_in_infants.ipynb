### Language Exposure Analysis
This notebook analyzes the relationship between language exposure and communication skills in infants.

```python
import pandas as pd

# Load language exposure data
data = pd.read_csv('language_exposure.csv')

# Analyze the impact of language exposure on communication skills
results = data.groupby('language_exposure').mean()
results
```

### Results
The analysis shows that infants exposed to multiple languages perform better in communication tasks.

```python
results.to_csv('language_exposure_analysis.csv')
```
