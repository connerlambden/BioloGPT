Import and preprocess RNA-seq datasets from wild tomato species; compare expression profiles of transcription factors like NAC29.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Download dataset from NCBI GEO using accession GSE288242
# Process expression data for selected transcription factors
# Plot expression levels over different species

df = pd.read_csv('gene_expression_data.csv')
selected_genes = df[df['gene'].isin(['NAC29', 'ethylene_response_factor'])]

plt.figure(figsize=(10,6))
for species in selected_genes['species'].unique():
    subset = selected_genes[selected_genes['species'] == species]
    plt.plot(subset['time_point'], subset['expression'], label=species)
plt.title('Expression Profiles of NAC29 Across Wild Tomato Species')
plt.xlabel('Time Point')
plt.ylabel('Expression Level')
plt.legend()
plt.show()
```

This analysis visualizes the temporal expression changes and facilitates cross-species comparisons of key transcription factors influencing QDR.
