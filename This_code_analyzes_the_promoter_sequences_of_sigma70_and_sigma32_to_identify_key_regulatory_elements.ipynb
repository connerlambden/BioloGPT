### Promoter Analysis of Sigma70 and Sigma32

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load promoter data
sigma70_promoters = pd.read_csv('sigma70_promoters.csv')
sigma32_promoters = pd.read_csv('sigma32_promoters.csv')

# Analyze and visualize the data
plt.figure(figsize=(10, 6))
plt.hist(sigma70_promoters['length'], alpha=0.5, label='Sigma70 Promoters')
plt.hist(sigma32_promoters['length'], alpha=0.5, label='Sigma32 Promoters')
plt.title('Length Distribution of Sigma70 and Sigma32 Promoters')
plt.xlabel('Length (bp)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
