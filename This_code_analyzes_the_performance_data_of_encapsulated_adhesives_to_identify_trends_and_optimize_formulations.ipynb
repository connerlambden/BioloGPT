### Data Analysis of Encapsulated Adhesives
This notebook will analyze the performance data of encapsulated adhesives, focusing on peel strength and environmental stability.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for encapsulated adhesives
data = {
    'Adhesive': ['MCs-Monomer', 'MCs-Trimer', 'MCs-Polymer'],
    '1_day': [67, 70, 69],
    '1_month': [56, 64, 54],
    '2_months': [42, '<10', 14],
    '3_months': ['<10', '<10', '<10']
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(df['Adhesive'], df['1_day'], marker='o', label='1 Day')
plt.plot(df['Adhesive'], df['1_month'], marker='o', label='1 Month')
plt.plot(df['Adhesive'], df['2_months'], marker='o', label='2 Months')
plt.plot(df['Adhesive'], df['3_months'], marker='o', label='3 Months')
plt.title('Performance of Encapsulated Adhesives Over Time')
plt.xlabel('Adhesive Type')
plt.ylabel('Isocyanate Loading (wt%)')
plt.legend()
plt.grid()
plt.show()
```
