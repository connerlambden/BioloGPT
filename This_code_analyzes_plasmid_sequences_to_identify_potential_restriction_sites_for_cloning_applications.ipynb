This notebook will analyze plasmid sequences to identify restriction enzyme sites, aiding in cloning strategy development.

```python
import Bio
from Bio import SeqIO

# Load plasmid sequences
sequences = SeqIO.parse('plasmids.fasta', 'fasta')

# Identify restriction sites
restriction_sites = {'EcoRI': 'GAATTC', 'BamHI': 'GGATCC'}

for seq in sequences:
    for enzyme, site in restriction_sites.items():
        if site in str(seq.seq):
            print(f'{enzyme} site found in {seq.id}')
```

This analysis will help in planning cloning experiments by identifying suitable restriction sites.
