This notebook will analyze genomic sequences from EV-C105 and other enteroviruses to identify key mutations.

```python
import pandas as pd
# Load genomic data
sequences = pd.read_csv('genomic_sequences.csv')
# Analyze mutations
mutations = sequences['mutation'].value_counts()
```

The analysis will provide insights into the frequency of specific mutations across different enterovirus strains.

```python
mutations.plot(kind='bar', title='Mutation Frequency in Enteroviruses')
```
