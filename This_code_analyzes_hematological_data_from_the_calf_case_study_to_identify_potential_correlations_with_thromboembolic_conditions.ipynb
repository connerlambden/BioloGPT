### Hematological Data Analysis
This section analyzes the hematological values of the calf to identify any significant deviations from normal ranges.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Hematological data
data = {
    'Variables': ['Erythrocytes', 'Hemoglobin', 'Hematocrit', 'WBC', 'Platelets', 'BUN', 'Creatinine', 'Alk. Phos.', 'AST', 'GGT', 'CK'],
    'Results': [794, 10.4, 30.5, 22.7, 164, 11.6, 0.3, 487, 263, 77, 661],
    'Reference Range': [760, 12.2, 33, 12, 63.7, 25, 0.9, 107, 37.4, 318.3, 12.1]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(df['Variables'], df['Results'], color='blue', label='Results')
plt.bar(df['Variables'], df['Reference Range'], color='red', alpha=0.5, label='Reference Range')
plt.title('Hematological Values vs Reference Ranges')
plt.xlabel('Variables')
plt.ylabel('Values')
plt.legend()
plt.show()
```
