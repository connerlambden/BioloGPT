This notebook analyzes microbial community data from soil samples treated with FeNPs under alkaline stress conditions.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load microbial community data
data = pd.read_csv('microbial_community_data.csv')

# Analyze diversity metrics
# Plot diversity indices
plt.figure(figsize=(10, 6))
plt.bar(data['treatment'], data['diversity_index'])
plt.title('Microbial Diversity Index by Treatment')
plt.xlabel('Treatment')
plt.ylabel('Diversity Index')
plt.show()
```

The analysis reveals how FeNP treatment affects microbial diversity compared to control samples.

```python
# Further analysis of functional genes
functional_data = pd.read_csv('functional_genes_data.csv')
# Plotting functional gene expression
plt.figure(figsize=(10, 6))
plt.plot(functional_data['gene'], functional_data['expression'], marker='o')
plt.title('Functional Gene Expression Under FeNP Treatment')
plt.xlabel('Gene')
plt.ylabel('Expression Level')
plt.xticks(rotation=45)
plt.show()
```
