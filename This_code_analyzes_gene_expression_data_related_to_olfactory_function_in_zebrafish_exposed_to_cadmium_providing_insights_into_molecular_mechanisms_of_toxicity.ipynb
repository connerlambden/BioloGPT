### Gene Expression Analysis of Zebrafish Exposed to Cadmium
This analysis focuses on the impact of cadmium on olfactory receptor gene expression in zebrafish.

```python
import pandas as pd
# Load gene expression data
gene_data = pd.read_csv('cadmium_exposure_genes.csv')
# Analyze changes in olfactory receptor expression
olfactory_genes = gene_data[gene_data['gene_type'] == 'olfactory']
olfactory_genes.describe()
```

### Results
The analysis reveals significant changes in the expression levels of olfactory receptor genes post-cadmium exposure.

```python
# Visualize gene expression changes
import matplotlib.pyplot as plt
olfactory_genes.plot(kind='bar', x='gene_name', y='expression_change')
plt.title('Olfactory Receptor Gene Expression Changes')
plt.show()
```
