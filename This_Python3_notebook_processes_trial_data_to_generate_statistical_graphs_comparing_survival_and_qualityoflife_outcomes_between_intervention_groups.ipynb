This notebook downloads patient-level trial data and produces a bar graph to compare survival rates between treatment groups.

```python
import pandas as pd
import plotly.express as px

# Sample data based on reported values
data = {'Group': ['Placebo', 'Dexmedetomidine'], 'Survival_Rate': [12.6, 15.5]}
df = pd.DataFrame(data)

fig = px.bar(df, x='Group', y='Survival_Rate', title='Overall Survival Rate Comparison', labels={'Survival_Rate': 'Survival Rate (%)'}, color='Group', color_discrete_map={'Placebo': '#FF5733', 'Dexmedetomidine': '#6A0C76'})
fig.show()
```
