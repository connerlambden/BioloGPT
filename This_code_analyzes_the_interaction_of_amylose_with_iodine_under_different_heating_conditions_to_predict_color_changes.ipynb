# Analysis of Amylose-Iodine Interaction
This notebook explores how heating affects the color of amylose when iodine is added.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Data on amylose and iodine interaction
# Hypothetical data for demonstration
heating_conditions = ['Unheated', 'Heated']
colors = ['Blue', 'Colorless']

# Create a DataFrame
data = pd.DataFrame({'Heating Condition': heating_conditions, 'Color': colors})

# Plotting the results
plt.figure(figsize=(8, 5))
plt.bar(data['Heating Condition'], data['Color'].apply(lambda x: 1 if x == 'Blue' else 0), color=['blue' if x == 'Blue' else 'gray' for x in data['Color']])
plt.title('Color Change of Amylose with Iodine under Heating Conditions')
plt.ylabel('Color Presence (1=Blue, 0=Colorless)')
plt.show()
```

This analysis demonstrates the expected color change of amylose when heated and iodine is added.
