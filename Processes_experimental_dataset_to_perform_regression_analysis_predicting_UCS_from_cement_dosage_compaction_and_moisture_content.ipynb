This notebook details data loading, regression modeling, and visualization steps using the paper's extracted dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Dataset from the paper
data = {'Cement_admixture': [5, 7, 9, 11, 13, 15],
        'UCS': [1.4, 1.52, 1.98, 2.63, 2.9, 3.15]}
df = pd.DataFrame(data)

# Linear Regression
coeffs = np.polyfit(df['Cement_admixture'], df['UCS'], 1)
model = np.poly1d(coeffs)

# Plotting
plt.scatter(df['Cement_admixture'], df['UCS'], color='purple')
plt.plot(df['Cement_admixture'], model(df['Cement_admixture']), color='green')
plt.title('Regression of UCS vs Cement Admixture')
plt.xlabel('Cement Admixture (%)')
plt.ylabel('UCS (MPa)')
plt.show()
```

The regression confirms the strong positive trend between cement dosage and UCS, reinforcing the paper's claims.
