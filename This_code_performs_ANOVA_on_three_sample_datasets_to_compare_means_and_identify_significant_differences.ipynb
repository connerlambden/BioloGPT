### ANOVA Analysis of Three Samples
This notebook will guide you through performing ANOVA on three sample datasets.

```python
import pandas as pd
import scipy.stats as stats

# Sample data
sample1 = [5, 6, 7, 8, 9]
sample2 = [6, 7, 8, 9, 10]
sample3 = [7, 8, 9, 10, 11]

# Perform ANOVA
f_statistic, p_value = stats.f_oneway(sample1, sample2, sample3)

print(f'F-statistic: {f_statistic}, p-value: {p_value}')
```

### Interpretation
If the p-value is less than 0.05, we reject the null hypothesis and conclude that at least one group mean is significantly different.
