This notebook describes data acquisition and processing of composite properties from nettle-PLA studies.

```python
import pandas as pd
import plotly.express as px
# Download dataset from provided repository URL
url = 'URL_0/dataset.csv'
df = pd.read_csv(url)
fig = px.line(df, x='Fiber_Content', y='Tensile_Strength', title='Tensile Strength vs Nettle Fiber Content')
fig.show()
```

The above code generates an interactive plot showing the trend of tensile strength with varying fiber loadings.

```python
# Additional analysis for thermal stability
fig2 = px.line(df, x='Fiber_Content', y='Thermal_Stability', title='Thermal Stability vs Nettle Fiber Content')
fig2.show()
```
