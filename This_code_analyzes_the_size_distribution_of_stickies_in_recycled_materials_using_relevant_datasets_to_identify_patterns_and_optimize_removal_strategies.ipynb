This notebook will analyze the size distribution of stickies in recycled materials.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load relevant datasets
dataset = pd.read_csv('stickies_data.csv')

# Analyze size distribution
size_distribution = dataset['size'].value_counts()

# Plotting the size distribution
plt.figure(figsize=(10,6))
plt.bar(size_distribution.index, size_distribution.values)
plt.title('Size Distribution of Stickies')
plt.xlabel('Size (mm)')
plt.ylabel('Frequency')
plt.show()
```
