This notebook loads calcium imaging datasets and computes edge angle dissimilarity metrics to assess the preservation of internal geometry across days.

```python
import numpy as np
import pandas as pd
# Load dataset (replace 'data_imaging.csv' with actual data file path)
data = pd.read_csv('data_imaging.csv')

def compute_edge_angle_dissimilarity(activity_matrix):
    angles = np.arctan2(activity_matrix[:,1], activity_matrix[:,0])
    sorted_angles = np.sort(angles)
    return np.mean(np.diff(sorted_angles))

result = compute_edge_angle_dissimilarity(data.values)
print('Mean edge angle dissimilarity:', result)
```

The code computes the mean difference between sorted angles, serving as a proxy for the stability of the internal neural geometry over time.
