This notebook processes experimental mating behavior data to calculate selection gradients for brain size across different habitat complexities.

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load dataset (replace 'habitat_complexity_data.csv' with actual dataset path)
data = pd.read_csv('data/habitat_complexity_data.csv')

def compute_gradient(subset):
    slope, intercept, r_value, p_value, std_err = stats.linregress(subset['brain_size'], subset['mating_success'])
    return slope, p_value

open_data = data[data['habitat'] == 'open']
complex_data = data[data['habitat'] == 'complex']

open_gradient, open_pvalue = compute_gradient(open_data)
complex_gradient, complex_pvalue = compute_gradient(complex_data)

print('Open Habitat - Gradient:', open_gradient, 'p-value:', open_pvalue)
print('Complex Habitat - Gradient:', complex_gradient, 'p-value:', complex_pvalue)
```

This analysis quantifies selection gradients, facilitating the comparison between habitat types using statistical inference.
