Step-by-step data retrieval and analysis of AEV01 effects on cancer cell viability and inflammatory marker expression.

```python
import pandas as pd
import numpy as np
import plotly.express as px

# Load experimental data
viability_data = pd.read_csv('path/to/viability_data.csv')
protein_data = pd.read_csv('path/to/protein_expression.csv')

# Calculate mean viability per dose
dose_response = viability_data.groupby('Dose').agg({'Viability': 'mean'}).reset_index()

# Plot dose-response curve
fig = px.line(dose_response, x='Dose', y='Viability', title='AEV01 Dose-Response Curve')
fig.show()

# Statistical analysis of protein markers
protein_summary = protein_data.groupby('Treatment').agg({'TP53': 'mean', 'CD36': 'mean'})
print(protein_summary)
```

The above analysis quantifies dose-dependent cytotoxic effects and molecular marker modulation, underpinning AEV01's mechanism.
