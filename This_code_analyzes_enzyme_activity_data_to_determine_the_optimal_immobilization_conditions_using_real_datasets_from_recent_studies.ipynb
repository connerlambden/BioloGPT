### Enzyme Activity Analysis
This notebook will analyze the activity of immobilized enzymes under various conditions.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load enzyme activity data
activity_data = pd.read_csv('enzyme_activity.csv')

# Analyze and visualize the data
plt.figure(figsize=(10, 6))
plt.plot(activity_data['Temperature'], activity_data['Activity'], marker='o')
plt.title('Enzyme Activity vs Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Activity (U)')
plt.grid()
plt.show()
```

### Discussion
This analysis helps identify the optimal temperature for enzyme activity, crucial for effective immobilization.
