```python
import plotly.graph_objects as go

# Sample data
farms = ['Zigong', 'Jintang', 'Shuangliu', 'Xinjin', 'Lezhi']
diarrheal = [16.00, 28.60, 33.33, 64.71, 37.50]
non_diarrheal = [6.25, 8.33, 20.00, 16.46, 18.18]

# Create traces for both sample types
trace1 = go.Bar(x=farms, y=diarrheal, name='Diarrheal Samples', marker_color='#6A0C76')
trace2 = go.Bar(x=farms, y=non_diarrheal, name='Non-Diarrheal Samples', marker_color='#A569BD')

# Layout settings
layout = go.Layout(title='Bopivirus B Positive Rates by Farm and Sample Type', barmode='group',
                   xaxis=dict(title='Farm Location'),
                   yaxis=dict(title='Positive Rate (%)'))

fig = go.Figure(data=[trace1, trace2], layout=layout)
fig.show()
```
