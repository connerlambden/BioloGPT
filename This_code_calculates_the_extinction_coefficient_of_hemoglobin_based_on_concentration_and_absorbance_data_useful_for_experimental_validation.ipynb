# Extinction Coefficient Calculation
This notebook calculates the extinction coefficient of hemoglobin using absorbance data at 405 nm.

```python
import numpy as np

# Sample data: absorbance values and concentrations
absorbance = np.array([0.5, 0.6, 0.7])  # Example absorbance values
concentration = np.array([0.001, 0.002, 0.003])  # Corresponding concentrations in M

# Calculate extinction coefficient
extinction_coefficient = absorbance / concentration
mean_extinction_coefficient = np.mean(extinction_coefficient)
mean_extinction_coefficient
```

The calculated mean extinction coefficient provides insights into hemoglobin's light absorption properties.

```python
# Output the mean extinction coefficient
mean_extinction_coefficient
```
