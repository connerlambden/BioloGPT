### Bacterial Motility Analysis
This notebook analyzes motility data from bacterial species to understand how hydrodynamic interactions influence spatial segregation.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load motility data
data = pd.read_csv('bacterial_motility.csv')

# Analyze and visualize the data
plt.scatter(data['speed'], data['persistence_length'])
plt.title('Bacterial Motility Characteristics')
plt.xlabel('Speed (µm/s)')
plt.ylabel('Persistence Length (µm)')
plt.show()
```

This analysis provides insights into how motility characteristics correlate with spatial segregation in bacterial communities.
