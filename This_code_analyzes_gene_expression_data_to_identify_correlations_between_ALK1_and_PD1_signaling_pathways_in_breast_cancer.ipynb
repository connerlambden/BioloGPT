# Analyze gene expression data for ALK1 and PD-1

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_data.csv')
sns.heatmap(data.corr(), annot=True)
plt.title('Correlation between ALK1 and PD-1 signaling genes')
plt.show()
```
