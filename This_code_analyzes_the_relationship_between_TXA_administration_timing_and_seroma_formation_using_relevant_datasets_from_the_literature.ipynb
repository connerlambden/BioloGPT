### Data Analysis of TXA and Seroma Formation
This notebook will analyze the impact of TXA timing on seroma formation using data from recent studies.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data based on literature
data = {
    'Group': ['TXA Preoperative', 'TXA Intraoperative', 'Control'],
    'Seroma Rate (%)': [5, 10, 15],
    'Drainage Volume (mL)': [40, 50, 60]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 5))
plt.bar(df['Group'], df['Seroma Rate (%)'], color='blue', alpha=0.7)
plt.title('Seroma Rates by TXA Administration Timing')
plt.ylabel('Seroma Rate (%)')
plt.show()
```
