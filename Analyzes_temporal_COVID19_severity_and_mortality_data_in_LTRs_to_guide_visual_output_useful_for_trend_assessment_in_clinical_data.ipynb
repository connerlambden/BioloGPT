This notebook downloads and processes COVID-19 clinical data in liver transplant patients, generating trend graphs.

```python
import pandas as pd
import plotly.express as px

data = {'Wave': ['Wave 1', 'Wave 2', 'Wave 3', 'Wave 4', 'Wave 5', 'Wave 6', 'Wave 7'], 'Severe_Rate': [45,50,55,57.14,53,48,50], 'Mortality_Rate': [10,12,15,21.43,18,14,16]}
df = pd.DataFrame(data)
fig = px.bar(df, x='Wave', y=['Severe_Rate','Mortality_Rate'], barmode='group', title='LTR COVID-19 Severity and Mortality')
fig.show()
```

The graph generated displays grouped bar charts of severe disease and mortality rates over the pandemic waves.
