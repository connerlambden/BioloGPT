This notebook details steps to load genomic data, run clustering on BGC profiles, and generate interactive network graphs.

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load the dataset containing BGC profiles and antifungal activity scores
bgc_data = pd.read_csv('path_to_bgc_data.csv')

# Construct a similarity network graph based on shared features
G = nx.Graph()
for index, row in bgc_data.iterrows():
    G.add_node(row['BGC_ID'], antifungal=row['Activity'])
# Example: Add edges based on similarity threshold
for i in range(len(bgc_data)):
    for j in range(i+1, len(bgc_data)):
        if abs(bgc_data.loc[i, 'Feature'] - bgc_data.loc[j, 'Feature']) < 0.5:
            G.add_edge(bgc_data.loc[i, 'BGC_ID'], bgc_data.loc[j, 'BGC_ID'])

plt.figure(figsize=(10, 10))
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=50, node_color='purple')
nx.draw_networkx_edges(G, pos, alpha=0.3)
plt.title('BGC Similarity Network', fontsize=16)
plt.show()
```

This code visualizes relationships among BGCs that are potentially linked to antifungal activity. It serves as a basis for further refining the correlations with functional assays.
