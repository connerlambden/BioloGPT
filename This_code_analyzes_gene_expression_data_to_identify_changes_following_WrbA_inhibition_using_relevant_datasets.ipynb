### Step 1: Load Required Libraries and Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load gene expression data
# Assuming data is in a CSV format for this example
gene_data = pd.read_csv('gene_expression_data.csv')
```

### Step 2: Data Preprocessing

```python
# Filter for genes of interest related to WrbA inhibition
genes_of_interest = ['wrbA', 'katE', 'malE']
filtered_data = gene_data[gene_data['gene'].isin(genes_of_interest)]
```

### Step 3: Analyze Expression Changes

```python
# Calculate fold changes and visualize
filtered_data['fold_change'] = filtered_data['expression_after'] / filtered_data['expression_before']
sns.barplot(x='gene', y='fold_change', data=filtered_data)
plt.title('Gene Expression Changes Following WrbA Inhibition')
plt.show()
```
