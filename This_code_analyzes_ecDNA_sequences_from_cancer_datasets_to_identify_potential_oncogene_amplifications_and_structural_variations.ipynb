This notebook will analyze ecDNA sequences to identify oncogene amplifications and structural variations.

```python
import pandas as pd
# Load ecDNA dataset
dataset = pd.read_csv('ecDNA_data.csv')
# Analyze oncogene amplifications
amplifications = dataset[dataset['copy_number'] > 2]
```

The analysis identifies oncogenes with high copy numbers, indicating potential targets for therapy.

```python
# Output results
amplifications.to_csv('high_copy_oncogenes.csv', index=False)
```
