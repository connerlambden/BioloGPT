Step 1: Load proteomics and miRNA datasets from HMEVs for infected and control groups.

```python
import pandas as pd
# Load the HMEV proteomics and miRNA data
proteomics = pd.read_csv('proteomics_data.csv')
mirna = pd.read_csv('mirna_data.csv')
# Merge datasets on common vesicle identifiers
data = pd.merge(proteomics, mirna, on='vesicle_id')
print(data.head())
```

Step 2: Perform pathway enrichment analysis to identify key functional networks.

```python
import statsmodels.api as sm
# Perform regression analysis to correlate expression levels with functional scores
X = data[['protein_expression', 'mirna_expression']]
y = data['functional_score']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```
