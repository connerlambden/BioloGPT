### Data Analysis of DAPI Assay Results
This notebook will analyze the DAPI assay results in relation to various AOPs.

```python
import pandas as pd

# Load DAPI assay data
dapi_data = pd.read_csv('dapi_assay_results.csv')

# Analyze correlations with AOPs
correlation_results = dapi_data.corr()
```

### Correlation Results
The correlation results will help identify significant relationships between DAPI assay outcomes and AOPs.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize correlations
sns.heatmap(correlation_results, annot=True)
plt.title('Correlation between DAPI Assay and AOPs')
plt.show()
```
