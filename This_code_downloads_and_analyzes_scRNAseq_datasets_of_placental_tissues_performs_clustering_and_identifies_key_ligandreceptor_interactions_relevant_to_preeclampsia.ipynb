Step 1: Load and preprocess the scRNA-seq data from GEO (accession GSE290578) to filter out low-quality cells and normalize expression levels.

```python
import scanpy as sc
adata = sc.read_10x_mtx('path_to_data')
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
```

Step 2: Perform clustering with Leiden algorithm and identify clusters corresponding to immune cell types.

```python
sc.pp.highly_variable_genes(adata, n_top_genes=2000)
sc.tl.pca(adata, svd_solver='arpack')
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
sc.tl.leiden(adata)
sc.pl.umap(adata, color=['leiden'])
```

Step 3: Analyze ligand-receptor interactions using CellPhoneDB or equivalent tools to pinpoint influencers like SPP1 and CD44.

```python
# Example placeholder for ligand-receptor analysis
import cellphonedb_analysis as cpdb
results = cpdb.run_analysis(adata, clusters='leiden')
print(results['significant_interactions'])
```
