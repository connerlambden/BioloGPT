This notebook downloads the SSR marker and phenotype data for Lithuanian potato cultivars, processes the resistance scores, and creates interactive visualizations.

```python
import pandas as pd
import plotly.graph_objects as go

# Load the data
data = pd.DataFrame({
    'Cultivar': ['VB Aista', 'Goda', 'VB Liepa', 'VB Meda', 'VB Rasa', 'VB Venta'],
    'Wart Resistance': [1.6, 3.2, 1.8, 1.6, 2.0, 1.8],
    'Late Blight Resistance': [7.2, 8.0, 8.0, 8.0, 7.4, 7.2]
})

# Create a grouped bar chart
fig = go.Figure(data=[
    go.Bar(name='Wart Resistance', x=data['Cultivar'], y=data['Wart Resistance'], marker_color='#6A0C76'),
    go.Bar(name='Late Blight Resistance', x=data['Cultivar'], y=data['Late Blight Resistance'], marker_color='#FFA500')
])

# Change the bar mode
fig.update_layout(barmode='group', title='Resistance Scores for Lithuanian Potato Cultivars', xaxis_title='Cultivar', yaxis_title='Resistance Score')
fig.show()
```

The above code generates a grouped bar chart that compares the resistance scores of each cultivar, facilitating quick insights into their performance for breeding purposes.
