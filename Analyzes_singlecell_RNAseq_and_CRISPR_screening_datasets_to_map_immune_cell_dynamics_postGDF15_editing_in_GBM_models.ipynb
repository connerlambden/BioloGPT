This notebook downloads and processes single-cell datasets from GBM models to evaluate immune cell shifts after CRISPR-mediated GDF15 editing.

```python
import scanpy as sc
adata = sc.read('path_to_gbm_scRNAseq_data.h5ad')
# Preprocess and cluster data
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.tl.pca(adata, svd_solver='arpack')
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
sc.tl.umap(adata)
sc.pl.umap(adata, color=['cell_type', 'GDF15'], save='umap_gdf15.png')
```

This analysis allows visualization of the immune landscape and the impact of GDF15 inactivation on cellular clusters.

```python
sc.tl.rank_genes_groups(adata, 'cell_type', method='t-test')
sc.pl.rank_genes_groups(adata, n_genes=10, sharey=False, save='rank_genes.png')
```
