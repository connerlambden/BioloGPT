Below is a step-by-step Jupyter Notebook outline for downloading the assembly metrics data and generating a bar graph visualization with Plotly.

```python
import plotly.graph_objs as go

# Define genome assembly metrics
metrics = [
    {'metric': 'Genome Size (Mb)', 'value': 211.46},
    {'metric': 'Contig N50 (Mb)', 'value': 5.04},
    {'metric': 'Scaffold N50 (Mb)', 'value': 5.39},
    {'metric': 'Protein-coding Genes', 'value': 14596},
    {'metric': 'Non-coding RNAs', 'value': 1673},
    {'metric': 'Repetitive Sequences (Mb)', 'value': 100.96}
]

# Prepare data for Plotly
metrics_names = [item['metric'] for item in metrics]
metrics_values = [item['value'] for item in metrics]

trace = go.Bar(x=metrics_names, y=metrics_values, marker=dict(color='#6A0C76'))
layout = go.Layout(title='Kappaphycus striatus Genome Assembly Metrics', xaxis=dict(title='Metrics'), yaxis=dict(title='Values'))
fig = go.Figure(data=[trace], layout=layout)
fig.show()
```
