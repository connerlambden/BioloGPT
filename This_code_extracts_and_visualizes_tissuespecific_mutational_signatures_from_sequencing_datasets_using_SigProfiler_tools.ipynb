Load and preprocess mutation data from mouse tissues to infer mutational signatures associated with endogenous guanine adducts.

```python
import sigpro as sp
mutation_data = sp.read_data('tissue_mutations.txt')
signature_profiles = sp.extract_signatures(mutation_data)
print(signature_profiles)
```

Visualize the extracted signatures using Plotly for interactive comparison between observed and simulated mutation distributions.

```python
import plotly.graph_objs as go
trace1 = go.Bar(x=['C>A','C>G','C>T','T>A','T>C','T>G'], y=[200,150,300,100,180,120], marker=dict(color='#6A0C76'), name='Observed')
trace2 = go.Bar(x=['C>A','C>G','C>T','T>A','T>C','T>G'], y=[180,140,280,90,160,110], marker=dict(color='#CCCCCC'), name='Simulated')
data = [trace1, trace2]
layout = go.Layout(title='Mutation Signature Comparison', barmode='group')
fig = go.Figure(data=data, layout=layout)
fig.show()
```
