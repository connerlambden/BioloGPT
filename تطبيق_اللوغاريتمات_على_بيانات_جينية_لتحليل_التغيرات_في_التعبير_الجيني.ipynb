تطبيق اللوغاريتمات على بيانات جينية لفهم التغيرات في التعبير الجيني.

```python
import pandas as pd
import numpy as np

data = pd.read_csv('gene_expression_data.csv')
log_data = np.log(data + 1)  # إضافة 1 لتجنب اللوغاريتم من الصفر.
```

البيانات الجينية تم تحويلها باستخدام اللوغاريتم الطبيعي لتحليل التغيرات.

```python
log_data.to_csv('log_transformed_data.csv', index=False)
```
