This notebook downloads the pertinent cohort data and uses Plotly to create a bar chart visualizing adjusted odds ratios with error bars for Substance Use Outcomes.

```python
import plotly.graph_objects as go

data = {'Substance': ['Tobacco', 'Other Drug', 'High-Risk Alcohol'], 'aOR': [2.1, 2.5, 11.5], 'CI': [[1.3, 3.3], [1.6, 4.1], [7.0, 18.6]]}

fig = go.Figure()
fig.add_trace(go.Bar(
    x=data['Substance'],
    y=data['aOR'],
    error_y=dict(
        type='data',
        array=[(ci[1]-ci[0])/2 for ci in data['CI']],
        visible=True
    ),
    marker_color='#6A0C76'
))

fig.update_layout(title='Adjusted Odds Ratios for Substance Use Outcomes', xaxis_title='Substance Category', yaxis_title='aOR')
fig.show()
```
