This notebook loads patient RNA-seq data from TCGA-LIHC and gepLiver, processes differential expression of FAM99B and DDX21, and visualizes the associated ribosome biogenesis pathways.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load TCGA and gepLiver expression data
tcga_data = pd.read_csv('path/to/TCGA_LIHC.csv')
gepliver_data = pd.read_csv('path/to/gepLiver.csv')
# Process data for FAM99B and DDX21
selected_genes = ['FAM99B', 'DDX21']
tcga_selected = tcga_data[selected_genes]
gepliver_selected = gepliver_data[selected_genes]
# Generate a boxplot to compare expression levels
plt.figure(figsize=(8,5))
plt.boxplot([tcga_selected['FAM99B'], gepliver_selected['FAM99B']], labels=['TCGA', 'gepLiver'])
plt.title('FAM99B Expression in HCC Cohorts')
plt.ylabel('Expression Level')
plt.show()
```

Next, computing correlation between FAM99B and DDX21 across samples to understand regulatory interplay.

```python
import numpy as np
correlation_tcga = np.corrcoef(tcga_selected['FAM99B'], tcga_selected['DDX21'])[0,1]
correlation_gepliver = np.corrcoef(gepliver_selected['FAM99B'], gepliver_selected['DDX21'])[0,1]
print('TCGA Correlation:', correlation_tcga)
print('gepLiver Correlation:', correlation_gepliver)
```

This analysis will help validate the inverse relationship between FAM99B expression and DDX21 levels as reported in the study.
