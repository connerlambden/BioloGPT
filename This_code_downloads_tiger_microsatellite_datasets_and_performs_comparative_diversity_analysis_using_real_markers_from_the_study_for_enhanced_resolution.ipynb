This section downloads and processes the microsatellite data from the provided studies to analyze allelic richness and heterozygosity trends over time.

```python
import pandas as pd
import numpy as np

# Download dataset (placeholder: actual dataset URL required)
df = pd.read_csv('https://biologpt.com/dataset/tiger_microsatellites.csv')

diversity_stats = df.groupby('Timepoint').agg({'Allelic_Richness': 'mean', 'Observed_Heterozygosity': 'mean'})
print(diversity_stats)

# Plotting the trends
import matplotlib.pyplot as plt

diversity_stats.plot(kind='bar', color=['#6A0C76', '#B565A7'], title='Genetic Diversity Trends in Amur Tigers')
plt.ylabel('Metric Value')
plt.show()
```

The above code computes basic genetic diversity statistics across sampled timepoints and visualizes the changes in allelic richness and observed heterozygosity.

```python
# Further analysis using scipy for statistical testing
from scipy import stats

# Example: Performing a t-test between early and later timepoints
early = df[df['Timepoint'] == 'Early']['Allelic_Richness']
later = df[df['Timepoint'] == 'Later']['Allelic_Richness']
t_stat, p_value = stats.ttest_ind(early, later)
print('T-statistic:', t_stat, 'P-value:', p_value)
```
