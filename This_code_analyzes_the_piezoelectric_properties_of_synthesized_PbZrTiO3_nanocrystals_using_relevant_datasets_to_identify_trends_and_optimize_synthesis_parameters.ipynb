### Data Analysis of PbZrTiO3 Nanocrystals

```python
import pandas as pd

# Load relevant datasets
dataset = pd.read_csv('path_to_dataset.csv')

# Analyze piezoelectric properties
properties = dataset[['Open-circuit voltage', 'Short-circuit current', 'Accuracy of motion detection']]
properties.describe()
```

### Summary of Analysis
This analysis provides insights into the performance metrics of the synthesized PbZrTiO3 nanocrystals.

```python
# Visualizing the performance metrics
import matplotlib.pyplot as plt

plt.bar(properties['Open-circuit voltage'], properties['Short-circuit current'])
plt.title('Performance Metrics of PbZrTiO3 Nanocrystals')
plt.xlabel('Open-circuit Voltage (V)')
plt.ylabel('Short-circuit Current (nA)')
plt.show()
```
