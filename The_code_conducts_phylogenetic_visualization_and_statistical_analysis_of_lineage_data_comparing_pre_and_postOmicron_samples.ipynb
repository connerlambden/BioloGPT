This notebook imports the extracted sample data to build visualizations that compare lineage prevalence across pandemic phases.

```python
import pandas as pd
import plotly.express as px

# Define the dataset based on the extracted data
data = [
    {'lineage': 'B.1.1.33', 'samples': 46, 'percentage': 52.9},
    {'lineage': 'B.1.1.28.2', 'samples': 15, 'percentage': 17.2},
    {'lineage': 'B.1.1.28.1', 'samples': 9, 'percentage': 10.3},
    {'lineage': 'BQ.1.1', 'samples': 60, 'percentage': 60.6},
    {'lineage': 'BE.9', 'samples': 9, 'percentage': 9.1},
    {'lineage': 'BQ.1.18', 'samples': 7, 'percentage': 7.1}
]

# Create dataframe
df = pd.DataFrame(data)

# Plotting sample counts grouped by phase (assumed grouping)
fig = px.bar(df, x='lineage', y='samples', text='percentage',
             title='SARS-CoV-2 Lineage Sample Distribution')
fig.update_traces(texttemplate='%{text}%', textposition='outside')
fig.show()
```

The above bar graph illustrates the comparative sample counts between pre-Omicron (B.1.1.x) and post-Omicron (BQ.1.1, BE.9, BQ.1.18) lineages, thus visually supporting the paper's conclusions.

```python
import numpy as np

# Adding a log-transformed scaling for further analysis
df['log_samples'] = np.log(df['samples'] + 1)
fig2 = px.line(df, x='lineage', y='log_samples', title='Log-Transformed Sample Counts')
fig2.show()
```
