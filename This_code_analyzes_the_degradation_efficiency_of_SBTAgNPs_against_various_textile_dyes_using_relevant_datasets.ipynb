### Analyzing Degradation Efficiency of SBT-AgNPs
This notebook will explore the degradation efficiencies of SBT-AgNPs against different textile dyes.

```python
import pandas as pd
import matplotlib.pyplot as plt

dye_data = {'Dye': ['Methylene Blue', 'Malachite Green', 'Azo Dye 1', 'Azo Dye 2'],
            'Degradation Efficiency (%)': [58.3, 60, 45, 50]}
dye_df = pd.DataFrame(dye_data)

dye_df.plot(x='Dye', y='Degradation Efficiency (%)', kind='bar', color='skyblue')
plt.title('Degradation Efficiency of SBT-AgNPs Against Textile Dyes')
plt.ylabel('Efficiency (%)')
plt.show()
```
