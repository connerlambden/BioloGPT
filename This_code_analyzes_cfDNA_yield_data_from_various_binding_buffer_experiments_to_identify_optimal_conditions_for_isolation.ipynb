# Analysis of cfDNA Yield Data
This notebook analyzes the yield of cfDNA based on different binding buffer compositions.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('cfDNA_yield_data.csv')
plt.figure(figsize=(10,6))
plt.bar(data['Buffer_Condition'], data['Yield'])
plt.xlabel('Buffer Condition')
plt.ylabel('cfDNA Yield (ng/mL)')
plt.title('cfDNA Yield by Binding Buffer Condition')
plt.show()
```
