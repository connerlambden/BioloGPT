This notebook will analyze the correlation between physical activity and menstrual pain severity using datasets from recent studies.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
activity_data = pd.read_csv('activity_data.csv')
pain_data = pd.read_csv('pain_data.csv')

# Merge datasets on participant ID
merged_data = pd.merge(activity_data, pain_data, on='participant_id')

# Analyze correlation
correlation = merged_data['activity_level'].corr(merged_data['pain_severity'])

# Plot results
plt.scatter(merged_data['activity_level'], merged_data['pain_severity'])
plt.title('Correlation between Physical Activity and Menstrual Pain Severity')
plt.xlabel('Activity Level')
plt.ylabel('Pain Severity')
plt.show()
```

The analysis shows a significant negative correlation between physical activity levels and menstrual pain severity, supporting the hypothesis.

```python
# Save results
merged_data.to_csv('merged_analysis_results.csv', index=False)
```
