### Step 1: Import Required Libraries
Import necessary libraries for data manipulation and analysis.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from ldsc import ldscore_regression
```

### Step 2: Load Aquaculture Datasets
Load relevant datasets from aquaculture studies to analyze genetic data.

```python
# Load datasets
salmon_data = pd.read_csv('path_to_salmon_data.csv')
oyster_data = pd.read_csv('path_to_oyster_data.csv')
```

### Step 3: Prepare Data for LD Score Regression
Prepare the data by filtering and formatting it for LD score regression analysis.

```python
# Data preparation
salmon_data_filtered = salmon_data[salmon_data['quality'] == 'high']
oyster_data_filtered = oyster_data[oyster_data['quality'] == 'high']
```

### Step 4: Run LD Score Regression
Execute the LD score regression analysis on the prepared datasets.

```python
results_salmon = ldscore_regression(salmon_data_filtered)
results_oyster = ldscore_regression(oyster_data_filtered)
```
