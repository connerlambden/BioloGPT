### Data Analysis of Kata Techniques Impact on Stereotypy Scores

```python
import pandas as pd
import scipy.stats as stats

# Data from the study
exercise_group = [12.53, 7.20, 8.07]
control_group = [14.47, 13.93, 13.40]

# Conducting t-tests
pre_post_exercise = stats.ttest_rel(exercise_group[0], exercise_group[1])
pre_post_control = stats.ttest_rel(control_group[0], control_group[1])

pre_post_exercise, pre_post_control
```
