This notebook will analyze GATA gene expression data from onions subjected to chromium and salt stress.

```python
import pandas as pd

data = pd.read_csv('GATA_expression_data.csv')

# Analyze expression levels
expression_summary = data.describe()
```

The summary statistics provide insights into the expression levels of GATA genes under different stress conditions.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(data['Gene'], data['Expression_Level'])
plt.title('GATA Gene Expression Levels')
plt.xlabel('GATA Genes')
plt.ylabel('Expression Level')
plt.show()
```

The bar chart visualizes the expression levels of GATA genes, highlighting those significantly induced under stress.
