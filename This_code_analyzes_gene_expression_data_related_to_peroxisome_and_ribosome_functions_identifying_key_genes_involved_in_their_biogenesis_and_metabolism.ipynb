### Gene Expression Analysis for Peroxisomes and Ribosomes

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_data.csv')
sns.boxplot(x='gene', y='expression', data=data)
plt.title('Gene Expression Levels')
plt.show()
```

This analysis visualizes the expression levels of genes associated with peroxisome and ribosome functions.

```python
# Example code to analyze differential expression
from scipy import stats

# Assuming 'data' is a DataFrame with gene expression data
results = stats.ttest_ind(data[data['condition'] == 'control']['expression'], data[data['condition'] == 'treatment']['expression'])
print('T-test results:', results)
```
