This notebook analyzes the relationship between microclimatic factors and yeast diversity in oak forests.

```python
import pandas as pd
# Load yeast diversity data
yeast_data = pd.read_csv('yeast_diversity.csv')
```

The following analysis will explore how temperature and moisture levels correlate with yeast species richness.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(data=yeast_data, x='Temperature', y='Species_Richness')
plt.title('Yeast Species Richness vs Temperature')
plt.show()
```
