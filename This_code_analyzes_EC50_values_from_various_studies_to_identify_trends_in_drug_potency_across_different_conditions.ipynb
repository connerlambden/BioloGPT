### EC50 Analysis Notebook
This notebook will analyze EC50 values from multiple studies to identify trends and correlations.

```python
import pandas as pd

# Load EC50 data from studies
data = pd.read_csv('ec50_data.csv')

# Analyze trends in EC50 values
trends = data.groupby('drug').mean()['EC50']
trends.sort_values(inplace=True)
trends
```
