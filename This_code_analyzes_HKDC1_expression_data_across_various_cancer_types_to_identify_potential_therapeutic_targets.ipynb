### Data Analysis of HKDC1 Expression in Cancer
This notebook will analyze HKDC1 expression levels in different cancer datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load HKDC1 expression data from relevant datasets
datasets = {'Liver Cancer': 'path_to_liver_cancer_data.csv', 'Colorectal Cancer': 'path_to_colorectal_cancer_data.csv'}

# Analyze and visualize the data
for cancer_type, path in datasets.items():
    data = pd.read_csv(path)
    plt.figure(figsize=(10, 6))
    plt.title(f'HKDC1 Expression in {cancer_type}')
    plt.bar(data['Sample'], data['Expression_Level'])
    plt.xlabel('Samples')
    plt.ylabel('Expression Level')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
