Perform exploratory data analysis with PCA to visualize clustering by fetal viral load.

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(expression_data.drop('Group', axis=1))
plt.figure(figsize=(8,6))
plt.scatter(pca_result[:,0], pca_result[:,1], c=expression_data['Group'].map({'High':1, 'Low':0}))
plt.title('PCA of Placental Gene Expression')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```
