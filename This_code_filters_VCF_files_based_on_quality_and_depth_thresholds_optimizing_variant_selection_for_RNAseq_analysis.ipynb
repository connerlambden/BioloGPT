### Step 1: Load Required Libraries
Import necessary libraries for handling VCF files.

```python
import pandas as pd
import vcf

# Load VCF file
vcf_reader = vcf.Reader(open('input.vcf', 'r'))
```

### Step 2: Define Filtering Criteria
Set the quality and depth thresholds for filtering.

```python
quality_threshold = 20
depth_threshold = 10

# Initialize a list to store filtered variants
filtered_variants = []
```

### Step 3: Filter Variants
Iterate through the VCF records and apply the filtering criteria.

```python
for record in vcf_reader:
    if record.QUAL >= quality_threshold and record.INFO['DP'] >= depth_threshold:
        filtered_variants.append(record)
```
