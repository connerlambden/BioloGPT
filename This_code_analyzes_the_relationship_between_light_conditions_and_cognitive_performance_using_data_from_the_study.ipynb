### Data Analysis of Light Conditions and Reaction Times

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from the study
light_conditions = ['SWL', 'LWL', 'BWL', 'DL']
mean_reaction_times = [304.20, 305.33, 309.21, 318.82]

# Create a DataFrame
data = pd.DataFrame({'Light Condition': light_conditions, 'Mean Reaction Time (ms)': mean_reaction_times})

# Plotting the data
plt.bar(data['Light Condition'], data['Mean Reaction Time (ms)'], color=['blue', 'red', 'white', 'grey'])
plt.title('Mean Reaction Time under Different Light Conditions')
plt.xlabel('Light Condition')
plt.ylabel('Mean Reaction Time (ms)')
plt.show()
```
