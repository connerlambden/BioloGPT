This notebook will analyze sensory data from consumer tests and correlate it with film properties.

```python
import pandas as pd

# Load sensory data
sensory_data = pd.read_csv('sensory_data.csv')

# Analyze correlations
correlations = sensory_data.corr()
```

The analysis will help identify which sensory attributes are most affected by CBAE concentrations.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize correlations
sns.heatmap(correlations, annot=True)
plt.title('Correlation between Sensory Attributes and CBAE Concentrations')
plt.show()
```
