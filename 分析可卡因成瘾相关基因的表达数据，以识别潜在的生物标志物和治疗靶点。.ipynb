本实验将使用RNA-seq数据分析可卡因成瘾相关基因的表达变化，识别关键基因和潜在的治疗靶点。

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('cocaine_addiction_genes.csv')
sns.heatmap(data.corr(), annot=True)
plt.title('Correlation Heatmap of Cocaine Addiction Genes')
plt.show()
```

通过热图分析基因之间的相关性，识别与可卡因成瘾相关的关键基因。

```python
# 进一步分析特定基因的表达变化
selected_genes = data[data['gene'].isin(['FOS', 'EGR1', 'IL6', 'JUN'])]
plt.plot(selected_genes['time'], selected_genes['expression'])
plt.title('Expression of Key Genes Over Time')
plt.xlabel('Time (days)')
plt.ylabel('Gene Expression Level')
plt.show()
```
