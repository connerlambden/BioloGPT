### Morphometric Analysis of Ceratomyxa Spores
This notebook analyzes the morphometric data of Ceratomyxa spores to explore their influence on dispersal mechanisms.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for Ceratomyxa spores
data = {
    'Measurement': ['Length (µm)', 'Thickness (µm)', 'Polar Capsule Length (µm)', 'Polar Capsule Width (µm)'],
    'Value': [8.6, 20.5, 3.5, 3.3]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Measurement'], df['Value'], color='skyblue')
plt.title('Morphometric Measurements of Ceratomyxa fialai')
plt.ylabel('Size (µm)')
plt.xticks(rotation=45)
plt.show()
```
