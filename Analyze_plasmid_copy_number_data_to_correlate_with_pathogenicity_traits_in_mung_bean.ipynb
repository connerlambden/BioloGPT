This notebook analyzes the relationship between plasmid copy number and pathogenicity traits in *Curtobacterium flaccumfaciens* isolates.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Isolate': ['CffT13932', 'BRIP 70623', 'Cff089'],
        'Plasmid_Copy_Number': [2.72, 2.38, 0],
        'Visual_Symptoms': [266.1, 127.5, 0],
        'Trifoliate_Dry_Weight': [520, 735, 1544]}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.bar(df['Isolate'], df['Plasmid_Copy_Number'], color='blue', label='Plasmid Copy Number')
plt.bar(df['Isolate'], df['Visual_Symptoms'], color='red', label='Visual Symptoms', alpha=0.5)
plt.bar(df['Isolate'], df['Trifoliate_Dry_Weight'], color='green', label='Trifoliate Dry Weight', alpha=0.5)
plt.title('Plasmid Copy Number and Pathogenicity Traits')
plt.xlabel('Isolate')
plt.ylabel('Values')
plt.legend()
plt.show()
```
