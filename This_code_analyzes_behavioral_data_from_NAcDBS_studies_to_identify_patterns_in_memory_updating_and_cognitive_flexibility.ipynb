### Data Analysis of NAc-DBS Effects
This notebook analyzes behavioral data from studies on NAc-DBS to explore its impact on memory updating and cognitive flexibility.

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('nac_dbs_study_data.csv')

# Analyze memory updating performance
data.groupby('stimulation_condition')['memory_updating_score'].mean()
```

### Results
The analysis will provide insights into how different stimulation conditions affect memory updating scores.

```python
# Visualize the results
import matplotlib.pyplot as plt

plt.bar(data['stimulation_condition'], data['memory_updating_score'])
plt.title('Memory Updating Scores by Stimulation Condition')
plt.xlabel('Stimulation Condition')
plt.ylabel('Memory Updating Score')
plt.show()
```
