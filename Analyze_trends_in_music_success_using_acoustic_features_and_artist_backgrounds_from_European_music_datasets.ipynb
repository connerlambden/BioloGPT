This notebook will analyze the relationship between acoustic features of songs and their success in European charts.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
music_data = pd.read_csv('european_music_data.csv')

# Analyze trends
success_trends = music_data.groupby('genre')['success_metric'].mean()

# Plot results
plt.figure(figsize=(10,6))
plt.bar(success_trends.index, success_trends.values)
plt.title('Average Success Metric by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Success Metric')
plt.show()
```

This analysis provides insights into which genres are currently thriving in the European music market.
