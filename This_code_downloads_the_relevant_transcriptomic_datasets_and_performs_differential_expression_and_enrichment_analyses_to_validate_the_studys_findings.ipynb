This section downloads transcriptomic datasets from GEO, preprocesses the data, and performs a differential expression analysis using Python modules such as pandas, numpy, and matplotlib for visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Download dataset using GEOquery equivalent in Python (e.g., GEOparse)
import GEOparse

gse = GEOparse.get_GEO(geo='GSE137570', destdir='.')
# Example: Extract expression data
expr_data = gse.pivot_samples('VALUE')

# Differential expression analysis placeholder
# Assume we have a metadata file 'metadata.csv'
metadata = pd.read_csv('metadata.csv')
# Join expression data with metadata and perform analysis
# This code is a template and should be refined with real dataset specifications

# Plot a heatmap of selected creatine-related genes
import seaborn as sns
creatine_genes = ['GATM', 'GAMT', 'SLC6A8', 'IGF1']
heatmap_data = expr_data.loc[creatine_genes]
plt.figure(figsize=(8,6))
sns.heatmap(heatmap_data, cmap='viridis')
plt.title('Heatmap of Creatine-Related Gene Expression')
plt.show()
```

The above code demonstrates how to process public gene expression data, isolate creatine-related genes, and visualize their expression patterns. Further enrichment analysis can be performed using packages like gseapy.

```python
# Additional code for enrichment analysis using gseapy
import gseapy as gp

# Assuming a gene list obtained from differential expression analysis
gene_list = creatine_genes

enr = gp.enrichr(gene_list=gene_list, organism='Human', gene_sets=['KEGG_2019_Human'], outdir='Enrichr_results')
print(enr.results.head())
```
