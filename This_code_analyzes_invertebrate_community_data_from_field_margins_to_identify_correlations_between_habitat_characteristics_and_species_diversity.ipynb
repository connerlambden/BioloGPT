### Invertebrate Community Analysis
This notebook analyzes the relationship between field margin characteristics and invertebrate diversity.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load invertebrate data
data = pd.read_csv('invertebrate_data.csv')

# Analyze correlation between habitat features and species richness
correlation = data.corr()

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Habitat Features and Invertebrate Diversity')
plt.show()
```
