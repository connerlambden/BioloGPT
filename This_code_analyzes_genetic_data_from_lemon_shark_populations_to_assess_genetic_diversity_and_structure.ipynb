### Genetic Analysis of Lemon Sharks
This notebook will analyze genetic data from lemon shark populations to understand their genetic diversity and structure.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('lemon_shark_genetic_data.csv')
sns.boxplot(x='population', y='genetic_diversity', data=data)
plt.title('Genetic Diversity of Lemon Shark Populations')
plt.show()
```
