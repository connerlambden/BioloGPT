### Patient Data Analysis for MRONJ Risk Factors

```python
import pandas as pd

# Load patient data
patient_data = pd.read_csv('patient_data.csv')

# Analyze risk factors
risk_factors = patient_data.groupby('medication').agg({'MRONJ': 'sum', 'total_patients': 'count'})
risk_factors['risk_percentage'] = (risk_factors['MRONJ'] / risk_factors['total_patients']) * 100
risk_factors.sort_values(by='risk_percentage', ascending=False)
```
