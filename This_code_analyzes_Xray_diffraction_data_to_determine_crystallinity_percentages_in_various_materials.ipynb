### Data Analysis of X-ray Diffraction Results
This section will analyze X-ray diffraction data to determine the crystallinity of different samples.

```python
import pandas as pd
import numpy as np

# Load XRD data
xrd_data = pd.read_csv('xrd_data.csv')

# Calculate crystallinity percentage
xrd_data['Crystallinity'] = (xrd_data['Crystalline_Peak_Height'] / (xrd_data['Crystalline_Peak_Height'] + xrd_data['Amorphous_Peak_Height'])) * 100

# Display results
xrd_data[['Sample', 'Crystallinity']]
```

### Results Interpretation
The calculated crystallinity percentages will help in understanding the material properties.

```python
# Displaying the results
print(xrd_data[['Sample', 'Crystallinity']])
```
