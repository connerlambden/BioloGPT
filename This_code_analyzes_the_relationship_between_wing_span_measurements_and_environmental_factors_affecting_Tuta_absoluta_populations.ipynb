### Data Analysis of Tuta absoluta Wing Span
This notebook will analyze the relationship between the wing span of Tuta absoluta and various environmental factors.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
wing_span_data = pd.read_csv('wing_span_data.csv')

# Analyze data
plt.scatter(wing_span_data['wing_span'], wing_span_data['pest_damage'])
plt.title('Wing Span vs Pest Damage')
plt.xlabel('Wing Span (mm)')
plt.ylabel('Pest Damage (%)')
plt.show()
```

This analysis will help understand how wing span influences pest damage in tomato crops.
