Below is a step-by-step Jupyter Notebook outline to load simulation data, compare polarization results between PolarBEAR and est-sfs, and generate comparative plots.

```python
import pandas as pd
import numpy as np
import plotly.express as px

# Download simulation data directly from a repository (URL placeholder)
data_url = 'https://gitlab.gwdg.de/molsysevol/polarizing_without_outgroup/raw/main/simulation_data.csv'
sim_data = pd.read_csv(data_url)
print(sim_data.head())
```

The dataset includes columns for genomic positions, ancestral state probabilities derived from PolarBEAR and est-sfs, and additional metrics for ARG quality.

```python
fig = px.scatter(sim_data, x='ARG_quality', y='PolarBEAR_accuracy',
                 color='Method',
                 title='PolarBEAR Accuracy vs ARG Quality',
                 labels={'ARG_quality':'ARG Reconstruction Quality', 'PolarBEAR_accuracy':'PolarBEAR Accuracy (%)'})
fig.show()
```

This plot helps evaluate how variations in ARG reconstruction quality impact the accuracy of SNP polarization, comparing PolarBEAR with established outgroup-based methods.

```python
summary_table = sim_data.groupby('Method').agg({'PolarBEAR_accuracy':'mean','estsfs_accuracy':'mean'}).reset_index()
print(summary_table)
# Display summary table using DataTables.js capabilities if integrated in a web context
```
