This section loads the sediment nutrient dataset and visualizes nutrient content differences between bays to assess the impact of bay isolation.

```python
import pandas as pd
import plotly.express as px

data = pd.DataFrame({
    'Bay': ['Östra Myttingeviken', 'Eke fjärd'],
    'C_org_content': [20, 6.54],
    'TN_content': [3, 0.11]
})

fig = px.bar(data, x='Bay', y=['C_org_content', 'TN_content'], barmode='group',
             title='Sediment Nutrient Content by Bay',
             labels={'value':'Nutrient Content (%)', 'variable':'Nutrient Type'})
fig.show()
```
