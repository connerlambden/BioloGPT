This notebook loads the human ISC organoid ATAC-seq and RNA-seq datasets, performs quality control, and carries out differential analysis using DESeq2 and HOMER.

```python
import os
import subprocess
# Code to download datasets and run quality control using FastQC
subprocess.run(['fastqc', 'input_file.fastq'])
# Differential expression analysis using DESeq2 would follow
```

Next, the notebook integrates chromatin accessibility data from ATAC-seq with gene expression changes from RNA-seq to identify regulatory elements altered in pouchitis-derived ISCs.

```python
import pandas as pd
# Load dataset
atac_data = pd.read_csv('atac_seq_results.csv')
rna_data = pd.read_csv('rna_seq_results.csv')
# Merge and analyze for differential regions
merged_data = pd.merge(atac_data, rna_data, on='gene_id')
print(merged_data.head())
```

The notebook further visualizes the differential binding sites and gene expression patterns using Plotly for interactive graphs.

```python
import plotly.express as px
fig = px.scatter(merged_data, x='log2FoldChange_ATAC', y='log2FoldChange_RNA', color='significance', title='ATAC vs RNA Differential Analysis')
fig.show()
```
