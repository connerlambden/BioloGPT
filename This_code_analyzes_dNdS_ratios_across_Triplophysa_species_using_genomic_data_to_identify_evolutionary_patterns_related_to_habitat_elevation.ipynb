### Data Preparation
Load genomic data for Triplophysa species and extract dN/dS ratios.

```python
import pandas as pd

data = pd.read_csv('triplophysa_dN_dS.csv')

# Display the first few rows of the dataset
data.head()
```

### Analysis of dN/dS Ratios
Calculate the correlation between dN/dS ratios and habitat elevations.

```python
correlation = data['dN/dS'].corr(data['elevation'])
print(f'Correlation between dN/dS and elevation: {correlation}')
```

### Visualization
Plot the relationship between dN/dS ratios and habitat elevations.

```python
import matplotlib.pyplot as plt

plt.scatter(data['elevation'], data['dN/dS'])
plt.title('dN/dS Ratios vs. Habitat Elevation')
plt.xlabel('Elevation (m)')
plt.ylabel('dN/dS Ratio')
plt.show()
```
