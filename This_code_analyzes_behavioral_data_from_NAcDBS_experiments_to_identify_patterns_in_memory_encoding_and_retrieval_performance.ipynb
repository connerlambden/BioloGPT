### Data Analysis of NAc-DBS Effects on Memory Encoding

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Condition': ['DBS-ON', 'DBS-OFF'], 'Performance': [0.75, 0.65]}
df = pd.DataFrame(data)

plt.bar(df['Condition'], df['Performance'], color=['blue', 'orange'])
plt.title('Memory Encoding Performance')
plt.ylabel('Performance (%)')
plt.show()
```

This analysis provides insights into the effectiveness of NAc-DBS in enhancing memory encoding.
