This notebook will explore the magnetic properties of iron filings and their alignment in magnetic fields.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated data for magnetic field strength and iron filing alignment
field_strength = np.linspace(0, 10, 100)
filing_alignment = np.sin(field_strength)  # Hypothetical alignment function

plt.plot(field_strength, filing_alignment)
plt.title('Iron Filing Alignment in Magnetic Field')
plt.xlabel('Magnetic Field Strength')
plt.ylabel('Filing Alignment')
plt.grid(True)
plt.show()
```

This graph illustrates how iron filings align in response to varying magnetic field strengths.
