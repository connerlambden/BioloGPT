### Genetic Cross Simulation
This notebook simulates the genetic cross between two individuals with specified genotypes to predict offspring phenotypes.

```python
# Define parent genotypes
parent1 = 'AA'  # Homozygous attached
parent2 = 'Aa'  # Heterozygous free

# Function to simulate Punnett square
import itertools

def punnett_square(parent1, parent2):
    offspring = list(itertools.product(parent1, parent2))
    return offspring

# Simulate the cross
results = punnett_square(parent1, parent2)
results
```

The results of the genetic cross will show the possible genotypes of the offspring.
