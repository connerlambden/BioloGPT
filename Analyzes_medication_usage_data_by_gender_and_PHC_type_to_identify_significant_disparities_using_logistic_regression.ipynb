### Analysis of Medication Usage by Gender and PHC Type
This analysis examines the distribution of antihypertensive medications among men and women across public and private primary health care centers.

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

### Data Preparation
Load and prepare the dataset for analysis.

```python
# Creating a dataframe with the extracted data
data = {
    'Medication': ['Three or more drugs', 'Calcium Channel Blockers', 'ARB', 'Beta Blockers', 'ACE Inhibitors', 'Thiazides'],
    'Men (%)': [60, 44, 43, 41, 36, 22],
    'Women (%)': [60, 44, 43, 41, 36, 22]  # Placeholder values, replace with actual data
}
df = pd.DataFrame(data)
```

### Statistical Analysis
Perform chi-squared tests to identify significant differences in medication usage between genders.

```python
# Example statistical test
for index, row in df.iterrows():
    chi2, p, dof, ex = stats.chi2_contingency([[row['Men (%)'], row['Women (%)']], [100-row['Men (%)'], 100-row['Women (%)']]])
    print(f"{row['Medication']}: p-value = {p}")
```

### Visualization
Visualize the differences in medication usage between genders.

```python
# Plotting the medication usage
plt.figure(figsize=(10,6))
sns.barplot(x='Medication', y='Men (%)', data=df, color='blue', label='Men')
sns.barplot(x='Medication', y='Women (%)', data=df, color='pink', label='Women')
plt.legend()
plt.title('Antihypertensive Medication Usage by Gender')
plt.show()
```
