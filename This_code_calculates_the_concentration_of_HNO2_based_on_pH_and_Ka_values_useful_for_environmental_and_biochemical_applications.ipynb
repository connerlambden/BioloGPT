### Calcolo della concentrazione di HNO2
Utilizzando i valori di pH e Ka, calcoliamo la concentrazione di HNO2.

```python
import numpy as np

# Dati
pH = 2.57
Ka = 7.1e-4

# Calcolo della concentrazione di ioni idrogeno
H_concentration = 10**(-pH)

# Risoluzione per C
C = (H_concentration**2) / Ka + H_concentration
C
```
