This notebook will analyze calcium transient data from non-excitable cells, focusing on pericytes and mast cells.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load calcium transient data
data = pd.read_csv('calcium_transients.csv')

# Analyze and visualize the data
plt.figure(figsize=(10, 6))
plt.plot(data['time'], data['calcium_level'])
plt.title('Calcium Transients in Non-Excitable Cells')
plt.xlabel('Time (s)')
plt.ylabel('Calcium Level (nM)')
plt.show()
```
