Este análisis se centra en la expresión de genes clave involucrados en la biogénesis ribosómica y su relación con el tamaño del nucleolo en células cancerosas.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_data.csv')
sns.boxplot(x='gene', y='expression', data=data)
plt.title('Expresión de genes relacionados con la biogénesis ribosómica')
plt.show()
```
