This notebook will analyze gene expression datasets related to CD8 T cell exhaustion, focusing on identifying key genes associated with reinvigoration.

```python
import pandas as pd
import numpy as np

# Load gene expression data
gene_data = pd.read_csv('gene_expression_data.csv')

# Analyze expression levels of key exhaustion markers
exhaustion_markers = ['PD-1', 'CTLA-4', 'LAG-3']
marker_expression = gene_data[gene_data['gene'].isin(exhaustion_markers)]

# Calculate average expression levels
average_expression = marker_expression.groupby('gene')['expression'].mean()
print(average_expression)
```
