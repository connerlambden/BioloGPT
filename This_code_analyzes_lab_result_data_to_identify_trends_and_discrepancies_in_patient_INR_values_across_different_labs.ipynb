### Lab Result Analysis
This notebook analyzes INR values from different labs to identify trends and discrepancies.

```python
import pandas as pd

# Sample data
lab_data = {
    'date': ['02/01/08', '08/22/08', '09/02/08', '09/06/08', '10/06/08'],
    'INR': [1.2, 2.9, 11.0, 2.4, 1.3],
    'Lab': ['C', 'A', 'B', 'A', 'C']
}

df = pd.DataFrame(lab_data)

# Analyze discrepancies
print(df.groupby('Lab')['INR'].describe())
```

### Analysis Results
The analysis provides insights into the INR values reported by different labs.

```python
# Display results
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.bar(df['Lab'], df['INR'], color='blue')
plt.title('INR Values by Lab')
plt.xlabel('Lab')
plt.ylabel('INR Value')
plt.show()
```
