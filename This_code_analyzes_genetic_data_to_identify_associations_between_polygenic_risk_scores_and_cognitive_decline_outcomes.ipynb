### Data Preparation
Load the HELIAD cohort data and extract relevant genetic and cognitive decline metrics.

```python
import pandas as pd

# Load data
cohort_data = pd.read_csv('HELIAD_cohort_data.csv')

# Extract relevant columns
relevant_data = cohort_data[['PRS', 'cognitive_decline', 'age', 'sex']]
```

### Statistical Analysis
Perform Cox proportional hazards analysis to evaluate the impact of PRS on cognitive decline.

```python
from lifelines import CoxPHFitter

# Fit the model
cph = CoxPHFitter()
cph.fit(relevant_data, duration_col='time', event_col='event')

# Display results
cph.print_summary()
```

### Visualization
Plot the results to visualize the relationship between PRS and cognitive decline.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(relevant_data['PRS'], relevant_data['cognitive_decline'])
plt.title('Cognitive Decline by Polygenic Risk Score')
plt.xlabel('Polygenic Risk Score')
plt.ylabel('Cognitive Decline')
plt.show()
```
