This notebook demonstrates the analysis of CRISPRi screening data alongside proteomics to identify key gene targets influencing isoprenol production.

```python
import pandas as pd
import plotly.express as px
data = pd.read_csv('crispri_data.csv')
result = data.groupby('gene')['yield'].mean().reset_index()
fig = px.bar(result, x='gene', y='yield', title='Gene Impact on Isoprenol Yield')
fig.show()
```

The above code loads the dataset, computes the average yield per gene, and visualizes the results interactively.
