### Home Range Analysis of Desert Birds
This notebook will analyze the home range sizes of various desert bird species and their correlation with body size.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for home range sizes and body sizes
data = {
    'Species': ['Savannah Sparrow', 'Cape Vulture (Adult)', 'Cape Vulture (Immature)'],
    'Home Range Size (ha)': [9.1, 11.0, 29.5],
    'Body Size (g)': [30, 7000, 8000]  # Example body sizes
}
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.scatter(df['Body Size (g)'], df['Home Range Size (ha)'], color='blue')
plt.title('Home Range Size vs Body Size in Desert Birds')
plt.xlabel('Body Size (g)')
plt.ylabel('Home Range Size (ha)')
plt.grid()
plt.show()
```
