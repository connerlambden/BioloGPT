Ova analiza koristi podatke o pulsu za istraživanje povezanosti između fizioloških i emocionalnih stanja.

```python
import pandas as pd

# Učitavanje podataka o pulsu
df = pd.read_csv('pulse_data.csv')

# Analiza varijabilnosti pulsa
df['pulse_variability'] = df['pulse'].std()  # Standardna devijacija pulsa
```

Ova analiza može pomoći u razumijevanju kako različiti faktori utječu na puls.

```python
# Vizualizacija rezultata
import matplotlib.pyplot as plt
plt.plot(df['time'], df['pulse_variability'])
plt.title('Varijabilnost pulsa tijekom vremena')
plt.xlabel('Vrijeme')
plt.ylabel('Varijabilnost pulsa')
plt.show()
```
