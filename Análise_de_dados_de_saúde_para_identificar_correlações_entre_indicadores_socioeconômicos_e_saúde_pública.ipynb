Análise de dados de mortalidade e morbidade em relação a indicadores socioeconômicos.

```python
import pandas as pd
# Carregar dados de saúde e socioeconômicos
dados = pd.read_csv('dados_saude.csv')
# Análise de correlação
correlacao = dados.corr()
```

Visualização dos resultados da análise de correlação.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(correlacao, annot=True)
plt.show()
```
