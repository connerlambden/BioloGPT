This notebook analyzes the impact of food insecurity on infant feeding practices using relevant datasets.

```python
import pandas as pd

# Load datasets
food_insecurity_data = pd.read_csv('food_insecurity_data.csv')
infant_feeding_data = pd.read_csv('infant_feeding_data.csv')

# Analyze correlations
correlation = food_insecurity_data.corrwith(infant_feeding_data['exclusive_breastfeeding'])
print(correlation)
```
