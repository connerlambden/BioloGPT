### Data Analysis of Neonatal Microbiome
This notebook analyzes sequencing data to identify key bacterial species influenced by delivery mode and diet.

```python
import pandas as pd
import plotly.express as px

# Load microbiome data
microbiome_data = pd.read_csv('path_to_microbiome_data.csv')
```

### Data Cleaning and Preparation
Ensure the data is clean and properly formatted for analysis.

```python
microbiome_data.dropna(inplace=True)
microbiome_data['Delivery_Mode'] = microbiome_data['Delivery_Mode'].astype('category')
microbiome_data['Diet'] = microbiome_data['Diet'].astype('category')
```

### Identify Key Bacterial Species
Analyze the abundance of key bacterial species across different delivery modes and diets.

```python
species = ['Lactobacillus', 'Bifidobacterium', 'Staphylococcus', 'Propionibacterium']
fig = px.box(microbiome_data, x='Delivery_Mode', y='Abundance', color='Delivery_Mode', facet_col='Species', category_orders={'Species': species})
fig.show()
```

### Conclusion
The analysis highlights significant differences in the abundance of specific bacterial species based on delivery mode and diet, corroborating the findings from Madan et al., 2012.
