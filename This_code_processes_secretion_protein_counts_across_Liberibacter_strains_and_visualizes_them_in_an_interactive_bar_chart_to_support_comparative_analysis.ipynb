This notebook downloads secretion protein count data for different Liberibacter strains and generates a Plotly bar chart.

```python
import plotly.graph_objs as go
import plotly.offline as pyo

data = [go.Bar(x=['Lso haplotype A', 'Lso haplotype B', 'Lso haplotype C', 'Lso haplotype D', 'CLas', 'L. crescens'], y=[102,75,85,77,66,69], marker=dict(color='#6A0C76'))]

layout = go.Layout(title='Non-classical Secreted Protein Counts in Liberibacter Genomes', xaxis=dict(title='Genome Strain'), yaxis=dict(title='Protein Count'))

fig = go.Figure(data=data, layout=layout)
pyo.plot(fig, filename='protein_count.html')
```
