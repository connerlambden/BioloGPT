This notebook retrieves quantitative vasculature parameters from high-content imaging and performs LDA to classify differences in phenotypic responses.

```python
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

data = pd.read_csv('imaging_data.csv')
features = data.iloc[:, :-1]
labels = data['group']
lda = LinearDiscriminantAnalysis(n_components=2)
transformed = lda.fit_transform(features, labels)
print(transformed.head())
```

The LDA results are visualized using Plotly to illustrate group separations.

```python
import plotly.express as px
fig = px.scatter(x=transformed[:,0], y=transformed[:,1], color=labels)
fig.show()
```
