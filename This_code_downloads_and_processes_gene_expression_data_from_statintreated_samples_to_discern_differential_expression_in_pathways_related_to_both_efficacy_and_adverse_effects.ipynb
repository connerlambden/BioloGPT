Load gene expression datasets related to statin treatment from BioStudies E-MTAB-3979 and perform differential analysis on pathways involved in muscle and metabolic responses.

```python
import pandas as pd
import numpy as np
from scipy import stats
# Download and process dataset (assume URL_ data available)
df = pd.read_csv('https://www.ebi.ac.uk/biostudies/files/E-MTAB-3979/E-MTAB-3979.json')
# Perform differential analysis
results = stats.ttest_ind(df[df['compound']=='simvastatin']['expression'], df[df['compound']=='control']['expression'])
print(results)
```

Visualize the differential expression results and identify candidate genes linked to adverse effects.

```python
import plotly.express as px
fig = px.histogram(df, x='expression', color='compound', title='Gene Expression Distribution')
fig.show()
```

The code helps identify key genes altered by statin treatment and supports hypothesis generation on adverse outcome mechanisms.
