This notebook will analyze gene expression data related to centrosome function and identify key genes associated with centrosome abnormalities.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('centrosome_expression_data.csv')
sns.boxplot(x='condition', y='expression_level', data=data)
plt.title('Centrosome Gene Expression Levels by Condition')
plt.show()
```
