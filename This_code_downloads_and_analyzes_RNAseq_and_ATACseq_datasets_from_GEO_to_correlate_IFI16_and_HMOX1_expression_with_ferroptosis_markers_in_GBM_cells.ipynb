Download the RNA-seq and ATAC-seq data from GSE274090 and preprocess using quality control and normalization.

```python
import GEOparse
# Download and parse the GEO series
geo = GEOparse.get_GEO(geo='GSE274090', destdir='./data')
# Process expression data
import pandas as pd
expr_data = geo.pivot_samples('VALUE')
print(expr_data.head())
```
