Overview: Import and inspect the lichen dataset, which includes historical and modern sections, to enable spatial and temporal analyses.

```python
import pandas as pd
# Load the dataset from the provided URL
url = 'https://cloud.gbif.org/eca/archive.do?r=to_cryptogamia-lichenes'
df = pd.read_csv(url)
print(df.head())
```

Next, process geolocation and collection date fields to prepare for Plotly visualizations.

```python
import plotly.express as px
fig = px.scatter(df, x='eventDate', y='decimalLatitude', color='stateProvince', title='Temporal and Spatial Distribution of Lichen Specimens')
fig.show()
```
