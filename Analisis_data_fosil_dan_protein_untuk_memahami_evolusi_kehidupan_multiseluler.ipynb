Analisis data fosil dan protein untuk memahami evolusi kehidupan multiseluler.

```python
# Mengimpor pustaka yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt

# Memuat data fosil
data = pd.read_csv('data_fosil.csv')

# Analisis data
print(data.describe())
```

Analisis ini memberikan wawasan tentang pola evolusi dan hubungan antara spesies.

```python
# Visualisasi data
plt.figure(figsize=(10,6))
plt.plot(data['waktu'], data['jumlah_spesies'], marker='o')
plt.title('Evolusi Spesies dari Waktu ke Waktu')
plt.xlabel('Waktu (juta tahun)')
plt.ylabel('Jumlah Spesies')
plt.show()
```
