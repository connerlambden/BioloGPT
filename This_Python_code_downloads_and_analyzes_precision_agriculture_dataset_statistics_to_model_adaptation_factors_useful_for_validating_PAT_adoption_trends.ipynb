This notebook details the process of downloading relevant datasets on PAT adoption, merging data sources, and performing statistical analyses to reveal correlations between socio-economic factors and technology adoption.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://biologpt.com/dataset/pat_adoption_v4.csv')
summary_stats = df.describe()
print(summary_stats)
```
