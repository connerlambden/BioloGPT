### Analyzing HLH Protein-DNA Interactions

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load data on HLH proteins and their binding sequences
data = pd.read_csv('hlh_proteins.csv')

# Analyze binding affinities
binding_affinities = data.groupby('protein')['affinity'].mean()
print(binding_affinities)
```
