Download and preprocess scRNAseq data from GSE283102, then perform differential expression analysis to identify key regulatory genes involved in the impaired function of tumor-induced memory CD8 T cells.

```python
import scanpy as sc
adata = sc.read_10x_mtx('path_to_GSE283102/')
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.tl.rank_genes_groups(adata, 'condition', method='t-test')
sc.pl.rank_genes_groups(adata, n_genes=20, sharey=False)
```

The above notebook provides a step-by-step differential expression workflow, essential for comparing memory CD8 T cell phenotypes.

```python
# Visualize clustering via UMAP
sc.pp.neighbors(adata, n_neighbors=10, n_pcs=40)
sc.tl.umap(adata)
sc.pl.umap(adata, color=['condition', 'gene_of_interest'])
```
