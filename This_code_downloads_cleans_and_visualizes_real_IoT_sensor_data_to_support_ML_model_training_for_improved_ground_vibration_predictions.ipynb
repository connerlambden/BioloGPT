This notebook demonstrates the workflow from data acquisition to model training using actual IoT sensor data from mine blasting operations.

```python
import pandas as pd
import plotly.express as px
# Download actual sensor data (URL placeholder)
data = pd.read_csv('http://datasource_link/sensor_data.csv')
print(data.head())
```

Next, we preprocess the data and extract key features for training.

```python
features = data[['distance_m', 'longitudinal_velocity_mm_s', 'transverse_velocity_mm_s', 'vertical_velocity_mm_s']]
target = data['peak_particle_velocity_mm_s']

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(features, target)
print('Model R2 score:', model.score(features, target))
```
