Load and preprocess the data of shark interactions and environmental measurements for statistical analysis and visualization.

```python
import pandas as pd
import plotly.express as px

# Load dataset
shark_data = pd.read_csv('shark_interactions.csv')
# Process data and create a bar chart for environmental influences
fig = px.bar(shark_data, x='Environmental Factor', y='Influence', color='Factor Category', title='Influence of Environmental Factors on Prey Discrimination')
fig.show()
```

The code above processes the dataset and creates an interactive Plotly bar chart to visualize the relative influence of each environmental factor.

```python
# Further analysis can include regression models and hypothesis tests to evaluate significance of each factor.
import statsmodels.api as sm

X = shark_data[['Tide Range', 'Visibility', 'Water Temperature', 'Sea Conditions']]
y = shark_data['Prey Choice']
X = sm.add_constant(X)
model = sm.Logit(y, X).fit()
print(model.summary())
```

The regression model provided helps in understanding the statistical significance of each environmental parameter in influencing prey discrimination.
