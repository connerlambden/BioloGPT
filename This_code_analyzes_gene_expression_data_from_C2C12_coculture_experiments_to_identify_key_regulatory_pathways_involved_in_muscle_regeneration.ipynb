This notebook will analyze gene expression data from C2C12 coculture experiments, focusing on immune cell interactions.

```python
import pandas as pd

# Load gene expression data
gene_data = pd.read_csv('gene_expression_data.csv')

# Analyze key pathways
pathway_analysis = gene_data.groupby('pathway').mean()
pathway_analysis.to_csv('pathway_analysis.csv')
```

The analysis identifies key pathways that are significantly altered in C2C12 cells when cocultured with immune cells.

```python
# Visualize results
import matplotlib.pyplot as plt
pathway_analysis.plot(kind='bar')
plt.title('Key Pathways in C2C12 Coculture')
plt.xlabel('Pathway')
plt.ylabel('Expression Level')
plt.show()
```
