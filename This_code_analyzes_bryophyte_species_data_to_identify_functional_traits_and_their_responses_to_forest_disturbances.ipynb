### Bryophyte Functional Traits Analysis
This notebook analyzes the functional traits of bryophyte species in relation to forest disturbances.

```python
import pandas as pd

# Load bryophyte species data
data = pd.read_csv('bryophyte_data.csv')

# Analyze functional traits
traits_summary = data.groupby('species').agg({'A_nmax_L': 'mean', 'SLA': 'mean'})
traits_summary
```

### Summary of Functional Traits
The analysis provides insights into how different bryophyte species adapt their functional traits in response to forest disturbances.

```python
# Display the summary
print(traits_summary)
```
