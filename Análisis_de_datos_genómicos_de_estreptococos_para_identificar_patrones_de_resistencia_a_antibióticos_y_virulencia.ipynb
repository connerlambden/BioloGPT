Análisis de datos de secuenciación de estreptococos para identificar genes de resistencia.

```python
import pandas as pd
# Cargar datos de resistencia a antibióticos
data = pd.read_csv('resistencia_estreptococos.csv')
# Análisis de patrones de resistencia
resistencia = data.groupby('genotipo')['resistencia'].mean()
```

Este análisis permite identificar qué cepas de estreptococos son más resistentes a los tratamientos.

```python
resistencia.to_csv('patrones_resistencia.csv')
```
