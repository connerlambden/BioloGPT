This notebook will analyze gene expression data from iPSC-derived pericytes to identify key markers and pathways involved in their differentiation.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('gene_expression_data.csv')
sns.boxplot(data=data, x='marker', y='expression')
plt.title('Gene Expression of Pericyte Markers')
plt.show()
```

The analysis reveals the expression levels of key pericyte markers, providing insights into the differentiation process.
