### Gene Expression Analysis for Ferroptosis Inducers

```python
import pandas as pd
import numpy as np

# Load gene expression data
expression_data = pd.read_csv('gene_expression_data.csv')

# Analyze expression levels of ABC transporters
abcb1_expression = expression_data['ABCB1']
abcg2_expression = expression_data['ABCG2']

# Calculate correlation with drug sensitivity
correlation = np.corrcoef(abcb1_expression, abcg2_expression)
print('Correlation between ABCB1 and ABCG2 expression:', correlation)
```
