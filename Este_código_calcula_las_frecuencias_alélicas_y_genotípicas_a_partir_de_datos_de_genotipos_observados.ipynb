### Cálculo de Frecuencias Alélicas y Genotípicas
Este código permite calcular las frecuencias alélicas y genotípicas a partir de un conjunto de datos de genotipos observados.

```python
# Importar bibliotecas necesarias
import numpy as np
import pandas as pd

# Datos de ejemplo: frecuencias de genotipos
# Supongamos que tenemos 200 individuos con los siguientes genotipos
# AA, Aa, aa
observed_genotypes = {'AA': 36, 'Aa': 48, 'aa': 64}

# Calcular frecuencias alélicas
N = sum(observed_genotypes.values())
p = observed_genotypes['AA'] + 0.5 * observed_genotypes['Aa']
q = observed_genotypes['aa'] + 0.5 * observed_genotypes['Aa']
p_freq = p / N
q_freq = q / N

# Calcular frecuencias genotípicas esperadas
p2 = p_freq ** 2
q2 = q_freq ** 2
pq = 2 * p_freq * q_freq

# Resultados
frequencies = {'p': p_freq, 'q': q_freq, 'p2': p2, 'pq': pq, 'q2': q2}
frequencies
```
