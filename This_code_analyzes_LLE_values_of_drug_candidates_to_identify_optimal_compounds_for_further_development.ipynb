### LLE Analysis of Drug Candidates
This notebook will analyze the LLE values of various drug candidates to identify those with the highest potential for success.

```python
import pandas as pd

# Load dataset containing pIC50 and logP values
drug_data = pd.read_csv('drug_candidates.csv')

drug_data['LLE'] = drug_data['pIC50'] - drug_data['logP']

# Display the top candidates based on LLE
top_candidates = drug_data.nlargest(10, 'LLE')
top_candidates
```

### Results
The top candidates based on LLE values are displayed above, indicating their potential for further development.

```python
# Visualizing the LLE values
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(top_candidates['Compound'], top_candidates['LLE'], color='blue')
plt.xlabel('Compound')
plt.ylabel('LLE Value')
plt.title('Top Drug Candidates by LLE')
plt.xticks(rotation=45)
plt.show()
```
