### Genetic Diversity Analysis of Lemna Species
This notebook analyzes genetic data from various Lemna populations to understand the impact of hybridization on adaptability.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load genetic diversity data
data = pd.read_csv('lemna_genetic_diversity.csv')

# Analyze genetic diversity
plt.figure(figsize=(10, 6))
plt.bar(data['species'], data['genetic_diversity'])
plt.title('Genetic Diversity in Lemna Species')
plt.xlabel('Species')
plt.ylabel('Genetic Diversity Index')
plt.show()
```

### Discussion
This analysis provides insights into how hybridization contributes to genetic diversity in Lemna species, enhancing their adaptability.
