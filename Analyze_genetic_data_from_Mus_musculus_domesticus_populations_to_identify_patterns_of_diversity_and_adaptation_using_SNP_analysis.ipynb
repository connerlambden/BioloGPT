This notebook will analyze SNP data from various populations of Mus musculus domesticus to assess genetic diversity and adaptation patterns.

```python
import pandas as pd
import numpy as np

# Load SNP data
snp_data = pd.read_csv('snp_data.csv')

# Analyze genetic diversity
diversity = snp_data.groupby('population').agg({'allele_freq': 'mean'})
print(diversity)
```
