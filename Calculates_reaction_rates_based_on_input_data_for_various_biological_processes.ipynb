This notebook calculates the reaction rate based on the amount of substance produced and the time taken.

```python
# Calculate reaction rate
def calculate_reaction_rate(amount_produced, time_taken):
    return amount_produced / time_taken

# Example usage
amount = 4  # grams
hours = 3  # hours
rate = calculate_reaction_rate(amount, hours)
rate
```

The above code defines a function to calculate the reaction rate and demonstrates its usage with the provided data.

```python
# Output the calculated rate
print(f'Reaction Rate: {rate:.2f} g/h')
```
