This notebook will analyze protein sequences to determine the frequency of amino acids following methionine.

```python
import pandas as pd

# Load protein sequences
sequences = pd.read_csv('protein_sequences.csv')

# Function to find amino acids following methionine

def find_following_amino_acids(sequences):
    following_aa = []
    for seq in sequences:
        if 'M' in seq:
            index = seq.index('M') + 1
            if index < len(seq):
                following_aa.append(seq[index])
    return following_aa

# Analyze sequences
following_amino_acids = find_following_amino_acids(sequences['sequence'])

# Count occurrences
count = pd.Series(following_amino_acids).value_counts()
count.to_csv('following_amino_acids_count.csv')
```

This analysis will provide insights into the most common amino acids that follow methionine in the provided protein sequences.

```python
# Display results
count.plot(kind='bar', title='Frequency of Amino Acids Following Methionine')
```
