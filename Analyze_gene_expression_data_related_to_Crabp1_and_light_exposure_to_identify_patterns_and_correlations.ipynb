This notebook will analyze gene expression data from Crabp1 neurons under different light conditions.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load gene expression data
data = pd.read_csv('crabp1_expression_data.csv')

# Visualize the data
sns.boxplot(x='light_condition', y='expression_level', data=data)
plt.title('Crabp1 Expression under Different Light Conditions')
plt.show()
```

The analysis visualizes how Crabp1 expression varies with light exposure, providing insights into its role in metabolic regulation.
