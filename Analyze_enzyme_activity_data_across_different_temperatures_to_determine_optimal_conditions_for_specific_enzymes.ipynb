This notebook will analyze enzyme activity data to identify optimal temperature ranges for various enzymes.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('enzyme_activity.csv')
plt.plot(data['Temperature'], data['Activity'])
plt.title('Enzyme Activity vs Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Activity')
plt.show()
```

The above code visualizes the relationship between temperature and enzyme activity, helping to identify optimal conditions.
