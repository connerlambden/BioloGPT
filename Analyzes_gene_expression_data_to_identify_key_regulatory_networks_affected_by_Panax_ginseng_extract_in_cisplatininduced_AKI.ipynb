## Gene Expression Analysis of Cisplatin-Induced AKI and Ginseng Treatment

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load gene expression data
data = pd.read_csv('gene_expression_data.csv')

display(data.head())
```

### Differential Expression of Cell Death-Associated Genes

```python
genes = ['p21', 'p27', 'p53', 'TIMP2', 'IGFBP7', 'NF-κB']
data_selected = data[data['gene'].isin(genes)]

plt.figure(figsize=(10,6))
sns.barplot(x='gene', y='expression_level', hue='group', data=data_selected)
plt.title('Gene Expression Levels in Different Treatment Groups')
plt.ylabel('Expression Level')
plt.xlabel('Genes')
plt.legend(title='Group')
plt.show()
```

### Correlation Analysis Between Serum Parameters and Gene Expression

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Serum Parameters and Gene Expression')
plt.show()
```

### Pathway Enrichment Analysis

```python
# Assuming pathway_data is loaded
from gseapy import enrichr

enr = enrichr(gene_list=genes, gene_sets='KEGG_2016', organism='Human', description='gene_sets')
enr.results.head()
```
