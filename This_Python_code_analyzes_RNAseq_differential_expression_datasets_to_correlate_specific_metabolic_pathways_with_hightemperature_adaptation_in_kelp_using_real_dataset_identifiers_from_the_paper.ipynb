Load and preprocess RNA-seq differential expression data from S japonica samples across temperature regimes.

```python
import pandas as pd
# Load kelp expression dataset from provided accession numbers
data = pd.read_csv('kelp_temperature_expression.csv')
df = data[data['condition'].isin(['high_temp', 'control'])]
print(df.head())
```

Perform pathway enrichment analysis using DESeq2 output to identify key metabolic pathways related to adaptation.

```python
# Example code block for pathway enrichment (pseudocode structure)
import gseapy as gp
# Enrichment analysis using gseapy
enr = gp.enrichr(gene_list=list(df['gene'][df['pvalue']<0.05]),
                 gene_sets='KEGG_2019_Human',
                 organism='Human',
                 description='kelp adaptation',
                 outdir='kelp_enrich',
                 cutoff=0.5)
print(enr.results.head())
```

Visualize the enriched pathways using Plotly.

```python
import plotly.express as px
fig = px.bar(enr.results, x='Term', y='Combined Score', title='Enriched Pathways in Kelp Adaptation')
fig.show()
```
