This notebook will analyze RNA-seq data from PDAC samples to evaluate the expression levels of immune-related genes following DNASE1L3 overexpression.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv('pdac_expression_data.csv')

# Filter for DNASE1L3 and immune-related genes
immune_genes = ['IL2', 'TNF', 'IFNG', 'CCL2', 'DNASE1L3']
filtered_data = dataset[dataset['gene'].isin(immune_genes)]

# Plot the expression levels
plt.figure(figsize=(10, 6))
sns.boxplot(x='gene', y='expression', data=filtered_data)
plt.title('Expression Levels of Immune-Related Genes in PDAC')
plt.show()
```

The results will provide insights into how DNASE1L3 overexpression influences the expression of key immune response genes.
