This notebook will analyze the resting membrane potential data from the study on GAL-101.

```python
import pandas as pd
import plotly.express as px

data = {'Condition': ['Control RGCs', 'Aβ1-42 RGCs', 'GAL-101 + Aβ1-42 RGCs', 'Control RPE', 'Aβ1-42 RPE', 'GAL-101 + Aβ1-42 RPE'],
        'RMP (mV)': [-68.3, -10, -50, -40.2, -10, -50]}
df = pd.DataFrame(data)
fig = px.bar(df, x='Condition', y='RMP (mV)', color='Condition', title='Resting Membrane Potential in RGCs and RPE Cells')
fig.show()
```
