# Locomotion Analysis of Crawfish and Starfish
This notebook analyzes the speed data of crawfish and starfish on various terrains.

```python
import pandas as pd

# Sample data for crawfish and starfish speeds
crawfish_data = {'Terrain': ['Rocky', 'Uneven', 'Smooth'], 'Speed_mps': [0.5, 0.4, 0.6]}
starfish_data = {'Terrain': ['Rocky', 'Uneven', 'Smooth'], 'Speed_mps': [0.1, 0.15, 0.2]}

crawfish_df = pd.DataFrame(crawfish_data)
starfish_df = pd.DataFrame(starfish_data)

# Merging data for comparison
comparison_df = pd.merge(crawfish_df, starfish_df, on='Terrain', suffixes=('_Crawfish', '_Starfish')
comparison_df
```

## Speed Comparison
This section will visualize the speed comparison between crawfish and starfish.

```python
import matplotlib.pyplot as plt

# Plotting the speed comparison
plt.figure(figsize=(10, 5))
plt.bar(comparison_df['Terrain'], comparison_df['Speed_mps_Crawfish'], color='blue', label='Crawfish Speed')
plt.bar(comparison_df['Terrain'], comparison_df['Speed_mps_Starfish'], color='orange', label='Starfish Speed', alpha=0.7)
plt.title('Speed Comparison of Crawfish and Starfish on Different Terrains')
plt.xlabel('Terrain')
plt.ylabel('Speed (m/s)')
plt.legend()
plt.show()
```
