### Text Summarization Analysis
This notebook will utilize the SMMRY and AutoSummarizer APIs to extract key sentences from provided text data.

```python
import requests

# Function to summarize text using SMMRY API
def summarize_text(text):
    response = requests.post('http://api.smmry.com/', data={'sm_api_input': text})
    return response.json()['sm_api_result']

# Example text
text = "Your long text goes here..."
summary = summarize_text(text)
print(summary)
```
