Load and preprocess clinical outcome data from the patient database.

```python
import pandas as pd
# Download dataset from provided URL and preprocess
url = 'https://data.example.com/clinical_outcomes.csv'
df = pd.read_csv(url)
df_clean = df.dropna()
df_clean.head()
```

Visualize correlation between pruritus intensity and quality of life measures.

```python
import plotly.express as px
fig = px.scatter(df_clean, x='Pruritus_Intensity', y='DLQI', color='Phenotype', title='Correlation between Pruritus Intensity and DLQI')
fig.show()
```
