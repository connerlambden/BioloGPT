Step-by-step analysis of DKD RNA-seq data focusing on CerS6 expression.

```python
import GEOparse
import pandas as pd
# Download dataset using GEOparse for accession number 'GSE184836'
gse = GEOparse.get_GEO(geo='GSE184836', destdir='.')
# Process and extract expression levels for CerS6
expr_data = gse.pivot_samples('VALUE')
cers6_expr = expr_data.loc[expr_data.index.str.contains('CerS6')]
print(cers6_expr.describe())
```

The above code downloads the transcriptomic data, pivots the samples, extracts CerS6 related expression, and prints summary statistics to help analyze its regulation in DKD.
