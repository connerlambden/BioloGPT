This notebook allows users to input genotype data and visualize the resulting phenotype ratios.

```python
import pandas as pd

def calculate_ratios(genotype_data):
    # Calculate genotype and phenotype ratios
    genotype_counts = genotype_data['Genotype'].value_counts()
    phenotype_counts = genotype_data['Phenotype'].value_counts()
    return genotype_counts, phenotype_counts
```

The function 'calculate_ratios' takes a DataFrame of genotype data and returns the counts of each genotype and phenotype.

```python
# Example usage
genotype_data = pd.DataFrame({'Genotype': ['GG', 'Gg', 'Gg', 'gg'], 'Phenotype': ['Dominant', 'Dominant', 'Dominant', 'Recessive']})
calculate_ratios(genotype_data)
```
