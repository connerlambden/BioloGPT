In this notebook, we load multi-sample spatial transcriptomics data processed by InSituPy. We utilize scanpy for single-cell data analysis and geopandas for spatial operations, demonstrating integration and visualization.

```python
import scanpy as sc
import geopandas as gpd
import pandas as pd

# Example: Load a pre-processed InSituPy AnnData object
adata = sc.read_h5ad('path_to_insitupy_data.h5ad')

# Basic preprocessing
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)

# Load corresponding spatial annotations
gdf = gpd.read_file('path_to_annotations.geojson')

# Integrate and visualize spatial distribution
sc.pl.spatial(adata, color='gene_of_interest')
print('Loaded and integrated spatial transcriptomics data successfully.')
```

The above code snippet demonstrates the integration of scanpy-based transcriptomic analysis with geopandas for spatial data handling, showcasing a typical workflow implementable within the InSituPy framework.
