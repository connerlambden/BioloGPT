### Data Loading and Preprocessing
This step loads gene expression data from AF and SLE patient cohorts for downstream analysis.

```python
import pandas as pd
# Load expression dataset
expression_data = pd.read_csv('af_sle_expression.csv')
print(expression_data.head())
```

### Differential Expression Analysis
Filter genes with p-values < 0.05 as significant DEGs.

```python
significant_genes = expression_data[expression_data['p_value'] < 0.05]
print(significant_genes[['gene', 'logFC']].head())
```

### Weighted Gene Co-expression Network Analysis (WGCNA)
Construct a network to identify key modules and hub genes.

```python
import networkx as nx
G = nx.Graph()
for idx, row in significant_genes.iterrows():
    G.add_node(row['gene'])
# (Edges based on correlation threshold would be added here)
print('Nodes in network:', G.nodes())
```

### Hub Gene Identification via LASSO Regression
Apply LASSO regression to select critical genes, focusing on TMEM45A.

```python
from sklearn.linear_model import Lasso
X = significant_genes[['gene_expression']].values
Y = significant_genes['clinical_score'].values
lasso = Lasso(alpha=0.01)
lasso.fit(X, Y)
hub_genes = [gene for gene, coef in zip(significant_genes['gene'], lasso.coef_) if coef != 0]
print('Identified Hub Genes:', hub_genes)
```

### Visualization
Plot TMEM45A expression levels across different patient groups.

```python
import plotly.express as px
fig = px.box(expression_data, x='group', y='TMEM45A_expression', title='TMEM45A Expression in AF vs Controls')
fig.show()
```
