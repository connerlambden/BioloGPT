### Lipid Storage Analysis of Bumblebee Queens
This analysis will focus on lipid storage data collected from various studies to understand the impact of diapause disruption.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for lipid storage
lipid_data = {'Condition': ['Control', 'Pesticide 6 ppb', 'Pesticide 60 ppb'],
              'Lipid Storage (µg/mg fat body)': [133.4, 115.1, 95.5]}

lipid_df = pd.DataFrame(lipid_data)

# Plotting the data
plt.bar(lipid_df['Condition'], lipid_df['Lipid Storage (µg/mg fat body)'], color=['green', 'orange', 'red'])
plt.title('Lipid Storage in Bumblebee Queens Under Different Conditions')
plt.ylabel('Lipid Storage (µg/mg fat body)')
plt.xlabel('Condition')
plt.show()
```
