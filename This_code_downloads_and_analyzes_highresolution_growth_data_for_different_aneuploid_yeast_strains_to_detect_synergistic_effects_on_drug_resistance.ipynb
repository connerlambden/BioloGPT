This notebook loads experimental growth datasets from engineered yeast strains and performs statistical analyses to identify significant gene dosage interactions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('path_to_growth_data.csv')
# Calculate normalized growth and plot synergistic interaction values
data['normalized_growth'] = data['growth_treated'] / data['growth_control']
plt.bar(data['strain'], data['normalized_growth'], color='#6A0C76')
plt.title('Growth Resistance Across Aneuploid Strains')
plt.xlabel('Strain')
plt.ylabel('Normalized Growth')
plt.show()
```

The above code demonstrates the computation of normalized growth values and visualizes the resistance phenotype across different yeast strains, helping to identify synergistic gene interactions.

```python
# Further statistical analysis could involve regression modeling to quantify interaction strengths
import statsmodels.api as sm
X = data[['chromosome1_effect', 'chromosome2_effect']]
X = sm.add_constant(X)
y = data['normalized_growth']
model = sm.OLS(y, X).fit()
print(model.summary())
```

This step-by-step analysis quantifies the contributions of individual and combined chromosomal alterations to the drug resistance phenotype.
