```python
import pandas as pd
import plotly.express as px

data = pd.read_csv('polyacetylene_data.csv')
fig = px.scatter(data, x='carbon_chain_length', y='bioactivity_index', color='structural_type', title='Polyacetylene Structure vs. Bioactivity')
fig.show()
```

The above code visualizes relationships between structural features and bioactivities, aiding chemotaxonomic insights.
