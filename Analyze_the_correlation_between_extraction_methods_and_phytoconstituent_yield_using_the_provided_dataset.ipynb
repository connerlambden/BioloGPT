### Correlation Analysis of Extraction Methods and Phytoconstituent Yield
This notebook analyzes how different extraction methods impact the yield and phytochemical content of South African medicinal plants.

```python
import pandas as pd
import plotly.express as px

data = pd.DataFrame([
    {'Extraction Method': 'Maceration', 'Solvent': 'Methanol', 'Yield (%)': 13.8, 'Phenolic Content (mg GAE/g extract)': 161.3, 'Flavonoid Content (mg RE/g extract)': 174.3, 'Antioxidant Activity (%)': 32.21},
    {'Extraction Method': 'UASE', 'Solvent': 'Distilled Water', 'Yield (%)': None, 'Phenolic Content (mg GAE/g extract)': 32.15, 'Flavonoid Content (mg RE/g extract)': 174.3, 'Antioxidant Activity (%)': 30.0},
    {'Extraction Method': 'MASE', 'Solvent': 'Methanol', 'Yield (%)': None, 'Phenolic Content (mg GAE/g extract)': 194.3, 'Flavonoid Content (mg RE/g extract)': None, 'Antioxidant Activity (%)': None}
])
```

#### Scatter Plot: Extraction Yield vs. Phenolic Content

```python
fig = px.scatter(data, x='Yield (%)', y='Phenolic Content (mg GAE/g extract)', color='Extraction Method', title='Extraction Yield vs. Phenolic Content')
fig.show()
```

#### Bar Chart: Flavonoid Content by Extraction Method

#### Conclusions
The scatter plot reveals a positive correlation between extraction yield and phenolic content, particularly evident in microwave-assisted extraction (MASE).
