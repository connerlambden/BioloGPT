### Optical Performance Analysis of Microlens Arrays
This notebook will analyze the optical performance data of microlens arrays, focusing on their transmission and reflection properties under different configurations.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load optical performance data
# Assuming data is in a CSV file with columns: 'Configuration', 'Transmission', 'Reflection'
data = pd.read_csv('microlens_performance.csv')

# Plotting the performance metrics
plt.figure(figsize=(10, 5))
plt.plot(data['Configuration'], data['Transmission'], label='Transmission (%)', marker='o')
plt.plot(data['Configuration'], data['Reflection'], label='Reflection (%)', marker='x')
plt.title('Optical Performance of Microlens Arrays')
plt.xlabel('Configuration')
plt.ylabel('Percentage (%)')
plt.legend()
plt.grid(True)
plt.show()
```
