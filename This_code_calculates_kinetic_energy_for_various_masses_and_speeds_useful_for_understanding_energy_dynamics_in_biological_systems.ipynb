### Kinetic Energy Calculation
This section calculates the kinetic energy for a range of masses and speeds.

```python
import numpy as np
import pandas as pd

# Define mass and speed ranges
masses = np.array([1, 2, 3, 4, 5])  # in kg
speeds = np.array([0, 1, 2, 3, 4, 5])  # in m/s

# Calculate kinetic energy
kinetic_energy = 0.5 * masses[:, np.newaxis] * speeds**2

# Create a DataFrame for better visualization
results = pd.DataFrame(kinetic_energy, index=masses, columns=speeds)
results
```
