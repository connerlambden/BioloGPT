### Ion Concentration Analysis
This section will analyze ion concentration data to predict membrane potential changes.

```python
import numpy as np

# Constants
R = 8.314  # J/(mol*K)
F = 96485  # C/mol
T = 310  # K (37 degrees Celsius)

# Function to calculate E_ion
def calculate_nernst_potential(C_out, C_in, z):
    return (R * T / (z * F)) * np.log(C_out / C_in)

# Example concentrations for K+
C_out_K = 5  # mM
C_in_K = 140  # mM
z_K = 1  # Valence for K+
E_K = calculate_nernst_potential(C_out_K, C_in_K, z_K)
E_K
```

### Results
The calculated equilibrium potential for K+ can be used to compare with the membrane potential.

```python
print(f'Equilibrium potential for K+: {E_K} mV')
```
