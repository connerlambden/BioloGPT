### Gene Expression Analysis of ODC in AD and DS
This notebook will analyze gene expression data to explore the relationship between ODC activation and AD progression in Down syndrome.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load gene expression data
data = pd.read_csv('gene_expression_data.csv')

# Analyze ODC expression levels
odc_expression = data[data['gene'] == 'ODC']
plt.figure(figsize=(10,6))
plt.plot(odc_expression['time'], odc_expression['expression'], label='ODC Expression')
plt.title('ODC Expression Over Time')
plt.xlabel('Time (months)')
plt.ylabel('Expression Level')
plt.legend()
plt.show()
```
