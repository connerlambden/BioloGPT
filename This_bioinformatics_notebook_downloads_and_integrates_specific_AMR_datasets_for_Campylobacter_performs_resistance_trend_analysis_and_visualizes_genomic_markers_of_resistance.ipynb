In this notebook, we will access and process the AMR datasets for Campylobacter isolates from Kigali. The code will extract resistance rates, construct statistical models, and produce interactive Plotly graphs.

```python
import pandas as pd
import plotly.express as px

# Download dataset (assuming URL exists in provided sources)
df = pd.read_csv('https://biologpt.com/data/campylobacter_amr_kigali.csv')

# Display key columns: prevalence, resistance rates
fig = px.bar(df, x='Antibiotic', y='Resistance_Rate', title='Antibiotic Resistance Rates in Campylobacter Isolates', color='Antibiotic', color_discrete_sequence=px.colors.qualitative.Plotly)
fig.show()
```

This visualization provides insights into the frequency of resistance across different antibiotics and assists in identifying high-risk antimicrobials.

```python
import statsmodels.api as sm

# Example: Linear regression to analyze relationship between sample size and resistance rate
df_reg = df[['Sample_Size', 'Resistance_Rate']]
df_reg = df_reg.dropna()
X = sm.add_constant(df_reg['Sample_Size'])
y = df_reg['Resistance_Rate']
model = sm.OLS(y, X).fit()
print(model.summary())
```

The regression analysis will help quantify the relationship between sample size and detected resistance, guiding further experimental design.
