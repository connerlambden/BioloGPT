Análisis de datos de cruces dihíbridos.

```python
import pandas as pd

# Datos de cruces dihíbridos
data = {'Fenotipo': ['Redonda Amarilla', 'Redonda Verde', 'Arrugada Amarilla', 'Arrugada Verde'],
        'Cantidad': [9, 3, 3, 1]}
df = pd.DataFrame(data)

# Cálculo de proporciones
proporciones = df['Cantidad'] / df['Cantidad'].sum()
proporciones
```

Este análisis permite visualizar las proporciones esperadas de los fenotipos en un cruce dihíbrido.

```python
# Visualización de resultados
import matplotlib.pyplot as plt

plt.bar(df['Fenotipo'], proporciones)
plt.title('Proporciones Fenotípicas en Cruce Dihíbrido')
plt.ylabel('Proporción')
plt.show()
```
