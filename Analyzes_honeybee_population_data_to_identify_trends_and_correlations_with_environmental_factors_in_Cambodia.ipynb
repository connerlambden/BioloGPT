# Honeybee Population Analysis in Cambodia
This notebook analyzes population data of native honeybees and correlates it with environmental factors such as habitat loss and pesticide usage.

```python
import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv('honeybee_population_cambodia.csv')

# Display the first few rows
df.head()
```

## Population Trends Over Time
Visualize the population trends of each native honeybee species over the years.

```python
fig = px.line(df, x='Year', y='Population', color='Species', title='Honeybee Population Trends in Cambodia')
fig.show()
```

## Correlation with Pesticide Usage
Analyze the correlation between pesticide usage and honeybee population declines.

```python
fig = px.scatter(df, x='Pesticide_Usage', y='Population', color='Species', trendline='ols', title='Correlation between Pesticide Usage and Honeybee Populations')
fig.show()
```
