This notebook analyzes the relationship between vaccination rates and pertussis outbreak incidence.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
vaccination_data = pd.read_csv('vaccination_rates.csv')
outbreak_data = pd.read_csv('pertussis_outbreaks.csv')

# Merge datasets
merged_data = pd.merge(vaccination_data, outbreak_data, on='year')

# Plotting
plt.figure(figsize=(10,6))
plt.scatter(merged_data['vaccination_rate'], merged_data['outbreak_cases'])
plt.title('Vaccination Rates vs Pertussis Outbreak Cases')
plt.xlabel('Vaccination Rate (%)')
plt.ylabel('Number of Outbreak Cases')
plt.show()
```

This analysis will help identify if there is a correlation between vaccination rates and the number of pertussis cases reported.

```python
# Further analysis can include regression models to predict outbreak likelihood based on vaccination rates.
```

This will provide insights into the effectiveness of current vaccination strategies.

```python
# Example regression analysis
import statsmodels.api as sm
X = merged_data['vaccination_rate']
y = merged_data['outbreak_cases']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
model.summary()
```
