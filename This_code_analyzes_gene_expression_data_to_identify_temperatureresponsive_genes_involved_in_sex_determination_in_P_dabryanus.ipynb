### Gene Expression Analysis for Temperature-Dependent Sex Determination in P. dabryanus

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load gene expression data
expression_data = pd.read_csv('gene_expression_data.csv')

# Filter for sex-determining genes
sex_determining_genes = ['sox3', 'cyp19a']
filtered_data = expression_data[expression_data['gene'].isin(sex_determining_genes)]

# Plot gene expression across temperatures
plt.figure(figsize=(10, 6))
sns.boxplot(x='temperature', y='expression', data=filtered_data)
plt.title('Gene Expression of Sex-Determining Genes at Different Temperatures')
plt.xlabel('Temperature (°C)')
plt.ylabel('Expression Level')
plt.show()
```

### Discussion
This analysis visualizes how temperature affects the expression of key sex-determining genes in P. dabryanus, providing insights into the molecular mechanisms of temperature-dependent sex determination.
