### SNP Inheritance Analysis
This notebook will analyze the inheritance of SNPs based on the provided parental genotypes.

```python
import pandas as pd

# Define the parental genotypes
mother_genotype = 'RV'
father_genotype = 'RR'
proband_genotype = 'RRR'

# Create a DataFrame to summarize the inheritance
data = {
    'Parent': ['Mother', 'Father', 'Proband'],
    'Genotype': [mother_genotype, father_genotype, proband_genotype],
    'Allele Contribution': ['1 R, 1 V', '2 R', '3 R (2 from father, 1 from mother)']
}

inheritance_df = pd.DataFrame(data)

# Display the DataFrame
inheritance_df
```

### Summary of Inheritance
The table above summarizes the contributions of each parent to the proband's genotype.

```python
# Display the DataFrame in a table format
inheritance_df.to_html('inheritance_table.html')
```
