### SNP Inheritance Analysis
This analysis will create a table to visualize the inheritance of alleles from parents to the proband.

```python
import pandas as pd

# Create a DataFrame for the SNP inheritance table
data = {
    'Individual': ['Mother', 'Father', 'Proband'],
    'Genotype': ['RV', 'RR', 'RRV'],
    'Alleles Contributed': ['R (1), V (1)', 'R (2)', 'R (3)']
}

snp_inheritance_df = pd.DataFrame(data)

# Display the DataFrame
snp_inheritance_df
```

### Summary of Results
The table above summarizes the contributions of each parent to the proband's genotype.

```python
# Display the table
snp_inheritance_df.to_html('snp_inheritance_table.html')
```
