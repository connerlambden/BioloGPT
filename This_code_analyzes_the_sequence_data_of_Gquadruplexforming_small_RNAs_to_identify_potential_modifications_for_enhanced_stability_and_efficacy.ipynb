### Sequence Analysis of G-Quadruplex-Forming Small RNAs
This section will analyze the sequences of G-quadruplex-forming small RNAs to identify potential modifications.

```python
import pandas as pd

# Load sequence data
sequences = pd.read_csv('g_quadruplex_sequences.csv')

# Analyze sequences for G-rich content
sequences['G_content'] = sequences['sequence'].apply(lambda x: x.count('G') / len(x))

# Display sequences with high G content
high_g_content = sequences[sequences['G_content'] > 0.6]
high_g_content
```

### Discussion
This analysis identifies sequences with high guanine content, which are potential candidates for G-quadruplex formation.
