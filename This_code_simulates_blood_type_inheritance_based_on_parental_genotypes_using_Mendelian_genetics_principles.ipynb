This notebook simulates the inheritance of blood types based on the ABO blood group system.

```python
# Simulate blood type inheritance
# Parent genotypes: Peeta (Aa), Katniss (BB)
parent1 = ['A', 'a']
parent2 = ['B', 'B']
children = [p1 + p2 for p1 in parent1 for p2 in parent2]
children
```

The output will show the possible blood types of the children.

```python
# Output the possible blood types
set(children) # This will show the unique blood types.
```

This simulation confirms that all children will have the AB blood type.

```python
# Final output
"Children's blood types: " + str(set(children))
```
