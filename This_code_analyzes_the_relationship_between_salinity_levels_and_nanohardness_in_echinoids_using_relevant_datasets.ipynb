### Data Analysis of Salinity and Nanohardness in Echinoids

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
# Assuming a dataset with columns 'salinity' and 'nanohardness'
data = pd.read_csv('echinoid_data.csv')

# Plotting the relationship
plt.scatter(data['salinity'], data['nanohardness'])
plt.title('Salinity vs Nanohardness in Echinoids')
plt.xlabel('Salinity (ppt)')
plt.ylabel('Nanohardness (GPa)')
plt.show()
```

This analysis visualizes the correlation between salinity levels and the nanohardness of echinoid skeletons, providing insights into their adaptive responses.
