### DNA Methylation Analysis of THC Exposure Data

```python
import pandas as pd
import numpy as np

# Load DNA methylation data
methylation_data = pd.read_csv('thc_methylation_data.csv')

# Identify differentially methylated regions (DMRs)
dmr_threshold = 0.05
significant_dmrs = methylation_data[methylation_data['p_value'] < dmr_threshold]

# Output significant DMRs
significant_dmrs.to_csv('significant_dmrs.csv', index=False)
```
