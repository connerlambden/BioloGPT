### Gene Expression Analysis of Protocadherins in Strabismus

```python
import pandas as pd

data = pd.read_csv('protocadherin_expression.csv')

# Analyze differential expression
results = data[data['expression_change'] > 1.5]
results.to_csv('upregulated_protocadherins.csv')
```

### Summary of Findings
This analysis identifies upregulated protocadherins that may be implicated in strabismus.

```python
# Visualize the results
import matplotlib.pyplot as plt
plt.bar(results['gene'], results['expression_change'])
plt.title('Upregulated Protocadherins in Strabismus')
plt.xlabel('Gene')
plt.ylabel('Expression Change')
plt.show()
```
