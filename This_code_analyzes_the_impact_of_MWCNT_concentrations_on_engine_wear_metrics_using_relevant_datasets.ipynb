### Data Analysis of MWCNT Effects on Engine Wear

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for analysis
data = {'Concentration': ['0 wt%', '0.5 wt%', '1 wt%', '1.5 wt%', '2 wt%'],
        'Friction Power (kW)': [2.3, 2.1, 1.8, 1.5, 1.25]}
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(df['Concentration'], df['Friction Power (kW)'], color='blue')
plt.title('Friction Power vs MWCNT Concentration')
plt.xlabel('MWCNT Concentration')
plt.ylabel('Friction Power (kW)')
plt.show()
```

### Discussion
This analysis shows the trend of friction power reduction with increasing MWCNT concentrations, indicating improved lubrication and reduced wear.
