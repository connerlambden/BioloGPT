This notebook will analyze the growth rates of different bacterial species on nutrient agar versus specialized media.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Bacteria': ['S.aureus', 'E.coli', 'H.influenzae'], 'Nutrient_Agar': [250, 200, 0], 'Chocolate_Agar': [300, 250, 150]}
df = pd.DataFrame(data)
df.set_index('Bacteria').plot(kind='bar')
plt.title('Bacterial Growth on Different Media')
plt.ylabel('Growth (CFU/ml)')
plt.show()
```

The above graph illustrates the differences in growth of various bacteria on nutrient agar compared to chocolate agar.
