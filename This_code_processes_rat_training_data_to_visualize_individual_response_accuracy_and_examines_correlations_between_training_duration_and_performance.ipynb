This notebook imports and visualizes rat behavioral data to identify trends in discrimination performance.

```python
import pandas as pd
import plotly.express as px

data = pd.DataFrame([
    {'rat': 'Darren', 'sessions': 37, 'correct': 87.2},
    {'rat': 'Lionel', 'sessions': 41, 'correct': 87.2},
    {'rat': 'Thomas', 'sessions': 64, 'correct': 88.9},
    {'rat': 'Peanuts', 'sessions': 67, 'correct': 82.8},
    {'rat': 'Schroedinger', 'sessions': 67, 'correct': 89.4},
    {'rat': 'Badger', 'sessions': 31, 'correct': 87.2},
    {'rat': 'Hero', 'sessions': 32, 'correct': 91.7},
    {'rat': 'Edison', 'sessions': 34, 'correct': 88.3},
    {'rat': 'Shaun', 'sessions': 67, 'correct': 89.4}
])
fig = px.bar(data, x='rat', y='correct', title='Rat Training Correct Response Percentages', labels={'correct':'Percentage Correct'})
fig.show()
```

The next cell calculates the correlation between the number of training sessions and response accuracy.

```python
import numpy as np
correlation = np.corrcoef(data['sessions'], data['correct'])[0,1]
print('Correlation between training sessions and accuracy:', correlation)
```
