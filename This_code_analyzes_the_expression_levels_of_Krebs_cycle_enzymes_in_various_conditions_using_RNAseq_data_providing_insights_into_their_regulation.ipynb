This notebook analyzes RNA-seq data to assess the expression of Krebs cycle enzymes under different metabolic conditions.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load RNA-seq data
data = pd.read_csv('krebs_cycle_expression.csv')

# Plot expression levels
plt.figure(figsize=(10,6))
plt.bar(data['Enzyme'], data['Expression_Level'])
plt.title('Expression Levels of Krebs Cycle Enzymes')
plt.xlabel('Enzyme')
plt.ylabel('Expression Level')
plt.xticks(rotation=45)
plt.show()
```

The plot above shows the expression levels of various Krebs cycle enzymes, highlighting their regulation under different conditions.
