This notebook will analyze RNA-seq data from CAR-T cells treated with MEK1/2 inhibitors and modified CD28 signaling.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('CAR_T_cell_data.csv')

# Analyze gene expression levels
sns.boxplot(x='treatment', y='gene_expression', data=data)
plt.title('Gene Expression in CAR-T Cells')
plt.show()
```
