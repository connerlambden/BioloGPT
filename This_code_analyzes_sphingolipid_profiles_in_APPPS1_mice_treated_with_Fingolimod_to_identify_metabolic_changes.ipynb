### Sphingolipid Analysis in APP/PS1 Mice
This notebook analyzes sphingolipid profiles from APP/PS1 mice treated with Fingolimod.

```python
import pandas as pd

# Load sphingolipid data
data = pd.read_csv('sphingolipid_data.csv')

# Analyze changes in sphingolipid levels
data.describe()
```

### Results
The analysis will provide insights into how Fingolimod treatment alters sphingolipid metabolism.

```python
# Visualize changes in sphingolipid levels
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.bar(data['sphingolipid'], data['level_change'])
plt.title('Sphingolipid Level Changes with Fingolimod Treatment')
plt.xlabel('Sphingolipid Species')
plt.ylabel('Change in Levels')
plt.show()
```
