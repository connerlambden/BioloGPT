### MNase-seq Data Analysis
This notebook will guide you through the analysis of MNase-seq data to identify nucleosome positions and accessibility.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load MNase-seq data
data = pd.read_csv('mnase_seq_data.csv')

# Analyze nucleosome positions
nucleosome_positions = data['position'].value_counts()

# Plot results
plt.figure(figsize=(10,6))
plt.bar(nucleosome_positions.index, nucleosome_positions.values)
plt.title('Nucleosome Positions from MNase-seq Data')
plt.xlabel('Position')
plt.ylabel('Frequency')
plt.show()
```

### Discussion
This analysis provides insights into nucleosome positioning across the genome, which is crucial for understanding chromatin dynamics.
