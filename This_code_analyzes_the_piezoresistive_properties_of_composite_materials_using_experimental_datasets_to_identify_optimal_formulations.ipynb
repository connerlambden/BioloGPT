### Piezoresistive Composite Analysis
This notebook analyzes the performance of various piezoresistive composites based on experimental data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
composite_data = pd.read_csv('composite_performance.csv')

# Analyze data
performance_summary = composite_data.describe()
performance_summary
```

### Summary of Performance
The summary provides insights into the mechanical and electrical performance of different composite formulations.

```python
# Visualize performance
plt.figure(figsize=(10, 6))
plt.bar(composite_data['formulation'], composite_data['sensitivity'])
plt.title('Piezoresistive Sensitivity of Composites')
plt.xlabel('Composite Formulation')
plt.ylabel('Sensitivity')
plt.show()
```
