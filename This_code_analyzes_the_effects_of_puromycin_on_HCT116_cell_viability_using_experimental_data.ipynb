# Analysis of Puromycin Effects on HCT116 Cells
This notebook will analyze the effects of different concentrations of puromycin on HCT116 cell viability.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for puromycin concentrations and cell viability
data = {'Concentration (µg/ml)': [0, 0.5, 1, 2], 'Viability (%)': [100, 90, 70, 30]}
viability_df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(8, 5))
plt.plot(viability_df['Concentration (µg/ml)'], viability_df['Viability (%)'], marker='o')
plt.title('Effect of Puromycin on HCT116 Cell Viability')
plt.xlabel('Puromycin Concentration (µg/ml)')
plt.ylabel('Cell Viability (%)')
plt.grid()
plt.show()
```
