This notebook will analyze genetic data from olive cultivars to identify markers associated with resistance to Olive Leaf Spot Disease.

```python
import pandas as pd
# Load genetic data
data = pd.read_csv('genetic_data.csv')
# Analyze markers associated with resistance
resistant_markers = data[data['resistance'] == 1]
```

The analysis identifies specific genetic markers that correlate with resistance to OLSD, providing insights for breeding programs.

```python
# Output the identified resistant markers
resistant_markers.to_csv('resistant_markers.csv', index=False)
```
