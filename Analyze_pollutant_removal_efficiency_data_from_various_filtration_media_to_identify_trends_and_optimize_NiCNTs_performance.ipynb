This notebook analyzes the pollutant removal efficiencies of Ni-CNTs compared to traditional materials.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Pollutant': ['CO2', 'CO', 'HC'], 'Ni-CNTs': [93.13, 94.87, 76.02], 'AC': [85.21, 80.77, 68.84], 'AC_MgO': [76.92, 86.84, 73.28]}
df = pd.DataFrame(data)
df.set_index('Pollutant').plot(kind='bar', figsize=(10, 6))
plt.title('Pollutant Removal Efficiencies')
plt.ylabel('Removal Efficiency (%)')
plt.show()
```

This analysis visualizes the comparative efficiencies of different filtration media.
