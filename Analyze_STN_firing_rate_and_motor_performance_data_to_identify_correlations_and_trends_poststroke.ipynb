### Data Analysis of STN Activity and Motor Performance Post-Stroke

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Condition': ['Before ACS', 'After ACS'],
        'STN_Firing_Rate': [69, 19],
        'Success_Rate': [74, 23]}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10,5))
plt.bar(df['Condition'], df['STN_Firing_Rate'], color='blue', alpha=0.6, label='STN Firing Rate')
plt.bar(df['Condition'], df['Success_Rate'], color='orange', alpha=0.6, label='Success Rate')
plt.title('STN Activity and Motor Performance Before and After Stroke')
plt.ylabel('Rate (%)')
plt.legend()
plt.show()
```
