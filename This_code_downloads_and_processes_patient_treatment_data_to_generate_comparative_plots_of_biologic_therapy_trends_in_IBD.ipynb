The notebook fetches real-world IBD treatment datasets and performs trend analysis over time.

```python
import pandas as pd
import plotly.express as px
df = pd.read_csv('https://data.source/ibd_treatment_trends.csv')
fig = px.bar(df, x='Therapy', y='Percentage', color='Year', barmode='group', title='IBD Biologic Trends')
fig.show()
```

Step-by-step instructions are provided to analyze the influence of patient demographics on therapy selection.

```python
import statsmodels.api as sm
X = df[['Age', 'Comorbidity_Score']]
y = df['Therapy_Choice']
model = sm.Logit(y, sm.add_constant(X)).fit()
print(model.summary())
```
