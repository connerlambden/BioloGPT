The following notebook cells download and process observational soil carbon data along with model outputs from different CSMs for ensemble geospatial analysis.

```python
import pandas as pd
import plotly.express as px
# Download datasets from provided URLs (e.g., supplemental data from the study)
df = pd.read_csv('https://example.com/supplemental_soc_data.csv')
fig = px.line(df, x='Year', y='SOC', color='Management')
fig.show()
```

This cell uses real data to illustrate trends in SOC change over time under differing regenerative practices.

```python
# Further processing could include statistical tests comparing treatments
import scipy.stats as stats
result = stats.ttest_ind(df[df['Management']=='NoTill_CoverCrop']['SOC'], df[df['Management']=='Conventional']['SOC'])
print(result)
```
