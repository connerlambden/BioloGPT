### GABPA Expression Analysis in Glioblastoma
This notebook analyzes GABPA expression levels in glioblastoma samples from TCGA and correlates them with TERT expression.

```python
import pandas as pd

gabpa_data = pd.read_csv('gabpa_expression_data.csv')
tert_data = pd.read_csv('tert_expression_data.csv')

merged_data = pd.merge(gabpa_data, tert_data, on='sample_id')
correlation = merged_data['gabpa_expression'].corr(merged_data['tert_expression'])
print(f'Correlation between GABPA and TERT expression: {correlation}')
```

### Results
The correlation coefficient indicates the strength of the relationship between GABPA and TERT expression in glioblastoma samples.

```python
# Visualizing the correlation
import matplotlib.pyplot as plt

plt.scatter(merged_data['gabpa_expression'], merged_data['tert_expression'])
plt.title('GABPA vs TERT Expression in Glioblastoma')
plt.xlabel('GABPA Expression')
plt.ylabel('TERT Expression')
plt.show()
```
