### Data Analysis of Insulin Aggregation Kinetics

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
lipid_concentrations = [0, 2, 4, 6, 8, 10]  # Example concentrations in µM
cytotoxicity = [100, 80, 60, 40, 20, 10]  # Corresponding cell viability percentages

# Create DataFrame
data = pd.DataFrame({'Lipid Concentration (µM)': lipid_concentrations, 'Cell Viability (%)': cytotoxicity})

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(data['Lipid Concentration (µM)'], data['Cell Viability (%)'], marker='o')
plt.title('Impact of Lipid Concentration on Insulin Cytotoxicity')
plt.xlabel('Lipid Concentration (µM)')
plt.ylabel('Cell Viability (%)')
plt.grid()
plt.show()
```
