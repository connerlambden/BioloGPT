Download and preprocess the spiking data from the GitHub repository to compute the firing rate changes across speaker ranks.

```python
import pandas as pd
import plotly.express as px

data = pd.read_csv('https://raw.githubusercontent.com/ccg1988/LSF_Nature_Communications_2024/main/Raw_data/data.csv')
fig = px.line(data, x='SpeakerRank', y='FiringRateChange', title='Spatial Modulation of Neuronal Responses')
fig.show()
```

The above code visualizes the relationship between speaker rank and neural firing rate changes, directly linking to the observed LSF phenomenon.
