### Step 1: Import Required Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2: Load nifH Sequence Data

```python
nifH_data = pd.read_csv('nifH_sequences.csv')  # Load your dataset here
```

### Step 3: Analyze Phylogenetic Relationships

```python
# Example analysis code
sns.countplot(data=nifH_data, x='Organism', hue='Phylogenetic_Group')
plt.title('Phylogenetic Distribution of nifH Sequences')
plt.xticks(rotation=45)
plt.show()
```
