This notebook will analyze sleep data from multiple studies to identify trends in compensatory sleep duration after sleep deprivation.

```python
import pandas as pd

# Load sleep data from studies
data = pd.read_csv('sleep_data.csv')

# Analyze compensatory sleep patterns
data['compensatory_sleep'] = data['post_deprivation_sleep'] - data['baseline_sleep']

# Summary statistics
summary = data.describe()
```

The analysis provides insights into how compensatory sleep varies across different studies and conditions.

```python
summary.to_csv('compensatory_sleep_summary.csv')
```
