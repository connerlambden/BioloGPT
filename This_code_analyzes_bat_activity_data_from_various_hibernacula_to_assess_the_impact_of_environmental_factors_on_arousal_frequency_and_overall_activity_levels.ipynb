### Data Analysis of Bat Activity in Hibernacula
This notebook will analyze bat activity data collected from different hibernacula, focusing on the influence of temperature and humidity on arousal frequency.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load bat activity data
data = pd.read_csv('bat_activity_data.csv')

# Analyze the relationship between temperature and activity levels
plt.scatter(data['temperature'], data['activity_level'])
plt.title('Bat Activity vs Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Activity Level')
plt.show()
```
