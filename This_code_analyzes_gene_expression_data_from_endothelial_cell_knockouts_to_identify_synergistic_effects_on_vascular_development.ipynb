This notebook will analyze gene expression data from endothelial cell knockouts to identify potential synergistic effects on vascular development.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('endothelial_knockout_data.csv')
sns.boxplot(x='gene', y='expression', data=data)
plt.title('Gene Expression in Endothelial Knockouts')
plt.show()
```
