In this notebook, we download MIC data for various Candida species exposed to anidulafungin from a continuous flow experiment dataset for visualization and comparative analysis.

```python
import pandas as pd
import plotly.express as px

# Sample dataset representing extracted MIC data
mic_data = pd.DataFrame({
    'Species': ['Candida albicans', 'Candida glabrata', 'Candida dubliniensis', 'Candida krusei', 'Candida lusitaniae', 'Candida tropicalis', 'Candida guilliermondii', 'Candida parapsilosis', 'Candida orthopsilosis', 'Candida metapsilosis'],
    'MIC_Range': ['0.003-0.016', '0.012-0.047', '0.016', '0.004', '0.023>32', '0.064->32', '>32', '1.5-3', '0.5-1.0', '0.25-0.3']
})

fig = px.bar(mic_data, x='Species', y=[1]*len(mic_data), text='MIC_Range', title='MIC Range Overview for Candida Species', labels={'y':'Relative Scale'})
fig.update_traces(textposition='outside')
fig.show()
```
