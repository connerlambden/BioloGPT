This notebook details the process of loading zebrafish metabolomic datasets, computing group statistics, and generating visualizations to highlight changes in GABA and glutamine levels.

```python
import pandas as pd
import plotly.express as px

data = pd.read_csv('zebrafish_metabolomics.csv')
fig = px.bar(data, x='metabolite', y='intensity', color='treatment', barmode='group', title='Metabolite Intensity by Treatment')
fig.show()
```

A statistical test comparing GABA levels in control versus nitrate-treated groups is then performed.

```python
import scipy.stats as stats

group_control = data[data['treatment']=='control']['GABA']
group_nitrate = data[data['treatment']=='nitrate']['GABA']
stat, p = stats.ttest_ind(group_control, group_nitrate)
print('GABA p-value:', p)
```
