### Data Analysis of Transformation Efficiency

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Fragment Size (mm)': [0.2, 0.5, 1.0, 2.0], 'Survival Rate (%)': [70, 100, 100, 100]}
df = pd.DataFrame(data)

plt.plot(df['Fragment Size (mm)'], df['Survival Rate (%)'], marker='o')
plt.title('Survival Rate of Chopped Thalli')
plt.xlabel('Fragment Size (mm)')
plt.ylabel('Survival Rate (%)')
plt.grid()
plt.show()
```

This analysis visualizes the relationship between fragment size and survival rate, aiding in optimizing transformation protocols.
