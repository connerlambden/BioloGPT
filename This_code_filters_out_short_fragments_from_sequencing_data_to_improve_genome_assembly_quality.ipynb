### Filtering Short Fragments from Sequencing Data
This notebook will demonstrate how to filter out fragments shorter than 100 base pairs from a given sequencing dataset.

```python
import pandas as pd

# Load sequencing data
sequencing_data = pd.read_csv('sequencing_data.csv')

# Filter out fragments shorter than 100 bp
filtered_data = sequencing_data[sequencing_data['length'] >= 100]

# Save the filtered data
filtered_data.to_csv('filtered_sequencing_data.csv', index=False)
```
