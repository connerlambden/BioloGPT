```python
import pandas as pd
# Download expression data
expression_data = pd.read_csv('https://www.example.com/dataset.csv')
# Preprocess and compare groups
grouped = expression_data.groupby('treatment').mean()
print(grouped.head())
```

This section outlines the step-by-step analysis for comparing expression profiles.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(data=expression_data, x='treatment', y='AMH_expression', palette='viridis')
plt.title('AMH Expression Levels by Treatment Group')
plt.show()
```

The plot above helps visualize differential expression of AMH across treatment groups.

```python
# Final statistical test
from scipy.stats import ttest_ind
control = expression_data[expression_data['treatment']=='control']['AMH_expression']
treated = expression_data[expression_data['treatment']=='treated']['AMH_expression']
stat, pvalue = ttest_ind(control, treated)
print('P-value:', pvalue)
```
