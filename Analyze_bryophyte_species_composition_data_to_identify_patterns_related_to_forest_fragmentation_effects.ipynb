This notebook analyzes bryophyte species composition data from fragmented and continuous forest sites.

```python
import pandas as pd

# Load bryophyte species data
bryophyte_data = pd.read_csv('bryophyte_species_data.csv')

# Analyze species richness and composition
richness = bryophyte_data.groupby('forest_type').size()

# Display results
richness
```
