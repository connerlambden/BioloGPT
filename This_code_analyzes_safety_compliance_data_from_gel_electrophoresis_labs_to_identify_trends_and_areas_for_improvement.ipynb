Analyze safety compliance data from gel electrophoresis labs.

```python
import pandas as pd

# Load safety compliance data
data = pd.read_csv('safety_compliance_data.csv')

# Analyze trends in compliance
compliance_trends = data.groupby('year')['compliance_rate'].mean()
print(compliance_trends)
```
