Este notebook está diseñado para analizar datos de experimentos químicos, utilizando bibliotecas como Pandas y Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data_experimento.csv')
plt.plot(data['Tiempo'], data['Concentracion'])
plt.title('Concentración a lo largo del tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Concentración (mol/L)')
plt.show()
```

El análisis de los datos permite identificar tendencias y patrones en los experimentos, facilitando la interpretación de los resultados.

```python
# Código para guardar la figura generada
plt.savefig('concentracion_tiempo.png')
```
