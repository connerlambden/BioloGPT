Below is a step-by-step Jupyter notebook snippet that processes BNCT experimental datasets to compare boron uptake and DNA damage in melanoma spheroids versus 2D cultures.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the dataset from the provided source URL
data = pd.read_csv('URL_for_dataset.csv')
# Group data by cell line and culture model for boron concentrations
uptake = data.groupby(['cell_line', 'model'])['boron_concentration'].mean().unstack()
uptake.plot(kind='bar', color=['#6A0C76', '#A0A0A0'])
plt.title('Average Boron Uptake in 3D vs 2D Models')
plt.xlabel('Cell Line')
plt.ylabel('Boron Concentration (µg 10B/10^6 cells)')
plt.show()
# Statistical comparison of DNA damage
from scipy.stats import ttest_ind
damage_3d = data[data['model'] == '3D']['DNA_damage']
damage_2d = data[data['model'] == '2D']['DNA_damage']
st, p_val = ttest_ind(damage_3d, damage_2d)
print('T-test p-value for DNA damage difference:', p_val)
```

This analysis helps to quantify whether the observed differences in boron uptake and DNA damage between 3D spheroids and 2D cultures are statistically significant.
