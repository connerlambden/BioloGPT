Este análisis se basa en datos de reacciones bioquímicas y sus energías libres de Gibbs.

```python
import pandas as pd

data = {'Reacción': ['Glu + ATP ↔ G6P + ADP', 'F6P + ATP ↔ FDP + ADP'], 'ΔG (kJ/mol)': [-16.74, -14.23]}
df = pd.DataFrame(data)
df.to_csv('gibbs_reactions.csv', index=False) # Guardar datos en un archivo CSV.
```

Los datos se pueden utilizar para realizar análisis adicionales sobre la espontaneidad de las reacciones.

```python
# Cargar datos y calcular estadísticas
import numpy as np

mean_deltaG = np.mean(df['ΔG (kJ/mol)'])
print(f'La energía libre de Gibbs promedio es: {mean_deltaG} kJ/mol')
```
