This notebook will analyze the accuracy of AI models on USMLE questions, comparing performance with and without images.

```python
import pandas as pd

# Load dataset
usmle_data = pd.read_csv('usmle_performance.csv')

# Analyze performance
performance_summary = usmle_data.groupby('Step').agg({'With_Images': 'mean', 'Without_Images': 'mean'})
performance_summary
```

The analysis shows the average accuracy of AI models on USMLE questions, highlighting the importance of visual data.

```python
performance_summary.to_csv('performance_summary.csv')
```
