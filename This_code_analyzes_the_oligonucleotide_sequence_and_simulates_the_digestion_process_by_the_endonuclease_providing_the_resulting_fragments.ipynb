### Oligonucleotide Digestion Simulation
This notebook simulates the digestion of a given oligonucleotide sequence by a specific endonuclease.

```python
# Define the oligonucleotide sequence
oligo_sequence = 'CCGGCCGTAACCGGCC'

# Define the recognition sequence and cleavage position
recognition_sequence = 'CCGGCC'
cleavage_position = 2  # Cleave after the second 'C'

# Function to simulate digestion
def digest_oligo(sequence, recognition, position):
    fragments = []
    if recognition in sequence:
        cut_index = sequence.index(recognition) + position + 1  # Position after the second 'C'
        fragment1 = sequence[:cut_index]
        fragment2 = sequence[cut_index:]
        fragments.append(fragment1)
        fragments.append(fragment2)
    return fragments

# Perform digestion
resulting_fragments = digest_oligo(oligo_sequence, recognition_sequence, cleavage_position)
resulting_fragments
```
