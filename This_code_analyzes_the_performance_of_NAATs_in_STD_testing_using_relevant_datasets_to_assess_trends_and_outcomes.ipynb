### NAAT Performance Analysis
This notebook will analyze the performance of NAATs in STD testing using data from public health laboratories.

```python
import pandas as pd

# Data for analysis
data = {
    'Test Type': ['Chlamydia', 'Gonorrhea'],
    'Total Tests': [3553196, 3461151],
    'NAAT Tests': [2289254, 2106055],
    'Non-NAAT Tests': [1263942, 1355096]
}
df = pd.DataFrame(data)

# Calculate NAAT percentage
df['NAAT Percentage'] = (df['NAAT Tests'] / df['Total Tests']) * 100

# Display the dataframe
df
```
