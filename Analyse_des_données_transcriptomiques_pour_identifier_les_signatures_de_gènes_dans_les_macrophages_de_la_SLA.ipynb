Analyse des données transcriptomiques des macrophages dans la SLA.

```python
import pandas as pd
# Charger les données des macrophages
macrophages_data = pd.read_csv('data_macrophages_sla.csv')
# Analyse des gènes exprimés
```

Visualisation des résultats.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.bar(macrophages_data['Gene'], macrophages_data['Expression'])
plt.title('Expression des gènes dans les macrophages de la SLA')
plt.xlabel('Gènes')
plt.ylabel('Niveau d’expression')
plt.show()
```
