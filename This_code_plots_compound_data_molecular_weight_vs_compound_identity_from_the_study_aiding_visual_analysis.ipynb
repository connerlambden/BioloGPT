### Data Visualization Notebook
This section downloads the compound data extracted from the paper and plots the molecular weights of dipyrimicin A and B using Plotly for interactive visualization.

```python
import plotly.graph_objects as go

# Define compound data
compounds = ['Dipyrimicin A', 'Dipyrimicin B']
molecular_weights = [247.0719, 246.0879]

# Create a bar chart
fig = go.Figure(data=[go.Bar(x=compounds, y=molecular_weights, marker_color='#6A0C76')])
fig.update_layout(title='Molecular Weights of Dipyrimicins', xaxis_title='Compound', yaxis_title='Molecular Weight', template='plotly_white')
fig.show()
```

### Analysis
The bar chart demonstrates the close molecular weights of the two compounds, providing a visual confirmation of their similar structural backbones.
