This notebook will analyze the relationship between feeding regimes, housing conditions, THI, and MMI using datasets from dairy sheep farms.

```python
import pandas as pd

# Load datasets
feeding_data = pd.read_csv('feeding_regimes.csv')
housing_data = pd.read_csv('housing_conditions.csv')
climate_data = pd.read_csv('climate_data.csv')

# Merge datasets based on farm ID
merged_data = feeding_data.merge(housing_data, on='farm_id').merge(climate_data, on='farm_id')

# Analyze the impact of feeding and housing on THI and MMI
results = merged_data.groupby(['feeding_regime', 'housing_condition']).agg({'THI': 'mean', 'MMI': 'mean'})
results.to_csv('analysis_results.csv')

# Display results
results
```

The analysis results will provide insights into how different feeding and housing strategies affect THI and MMI, guiding future management decisions.

```python
# Displaying the results
import matplotlib.pyplot as plt
results.plot(kind='bar')
plt.title('Impact of Feeding and Housing on THI and MMI')
plt.xlabel('Feeding Regime and Housing Condition')
plt.ylabel('Mean THI and MMI')
plt.show()
```
