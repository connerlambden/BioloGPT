### Data Analysis of Bat Echolocation Calls
This notebook will analyze echolocation call data from different bat species in Saudi Arabia.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load echolocation data
data = pd.read_csv('bat_echolocation_data.csv')

# Analyze call frequency and duration
plt.figure(figsize=(10, 6))
plt.scatter(data['Peak Frequency (kHz)'], data['Duration (ms)'])
plt.title('Echolocation Call Frequency vs Duration')
plt.xlabel('Peak Frequency (kHz)')
plt.ylabel('Duration (ms)')
plt.grid(True)
plt.show()
```

This analysis will help identify trends in echolocation calls among different bat species.
