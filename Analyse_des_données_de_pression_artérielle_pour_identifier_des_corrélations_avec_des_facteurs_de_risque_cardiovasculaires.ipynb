Analyse des données de pression artérielle et des facteurs de risque.

```python
import pandas as pd

# Charger les données de pression artérielle
data = pd.read_csv('blood_pressure_data.csv')

# Analyser les corrélations
correlations = data.corr()
```

Visualisation des corrélations entre la pression artérielle et les facteurs de risque.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualiser les corrélations
sns.heatmap(correlations, annot=True)
plt.title('Corrélations entre la pression artérielle et les facteurs de risque')
plt.show()
```
