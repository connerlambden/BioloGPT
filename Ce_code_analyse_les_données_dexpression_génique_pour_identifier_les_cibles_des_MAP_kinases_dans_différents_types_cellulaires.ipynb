Analyse des données d'expression génique pour identifier les cibles des MAP kinases.

```python
import pandas as pd
# Charger les données d'expression génique
data = pd.read_csv('gene_expression_data.csv')
# Analyser les cibles des MAP kinases
mapk_targets = data[data['gene'].isin(['ERK', 'JNK', 'p38'])]
```

Cette analyse permet d'identifier les gènes régulés par les MAP kinases dans les échantillons.

```python
mapk_targets.to_csv('mapk_targets.csv', index=False)
```
