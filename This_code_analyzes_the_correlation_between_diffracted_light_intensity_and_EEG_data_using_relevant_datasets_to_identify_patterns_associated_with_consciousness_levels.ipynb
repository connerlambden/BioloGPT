This notebook will analyze the correlation between diffracted light intensity and EEG markers of consciousness during anesthesia transitions.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
light_data = pd.read_csv('light_intensity_data.csv')
eeg_data = pd.read_csv('eeg_data.csv')

# Merge datasets on common identifiers
merged_data = pd.merge(light_data, eeg_data, on='subject_id')

# Calculate correlation
correlation = merged_data.corr()

# Visualize correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Light Intensity and EEG Markers')
plt.show()
```
