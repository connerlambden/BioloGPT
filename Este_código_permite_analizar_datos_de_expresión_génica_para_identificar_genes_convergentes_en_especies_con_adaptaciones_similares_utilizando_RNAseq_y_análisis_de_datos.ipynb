### Análisis de Datos de Expresión Génica
Este análisis se centrará en la identificación de genes convergentes en especies con adaptaciones similares. Utilizaremos datos de RNA-seq para este propósito.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar datos de expresión génica
expression_data = pd.read_csv('data/expression_data.csv')

# Análisis de expresión diferencial
# Filtrar genes con cambios significativos en la expresión
significant_genes = expression_data[expression_data['p_value'] < 0.05]

# Visualizar resultados
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='expression_level', data=significant_genes)
plt.title('Expresión Génica Significativa entre Especies')
plt.show()
```

### Discusión
Los resultados de este análisis ayudarán a identificar genes que son potencialmente responsables de adaptaciones convergentes en las especies estudiadas.

```python
# Guardar resultados significativos
significant_genes.to_csv('data/significant_genes.csv', index=False)
```
