This notebook downloads and visualizes the experimental inhibitor data for TOP3B and TOP1, enhancing comparative analysis.

```python
import plotly.graph_objects as go

compounds = ['bemcentinib', '2710-31', '2710-33-343', '2710-33-467', '2710-34-357', '2710-34-481']
top3b_ic50 = [34.3, 29.6, 29.3, 45.2, 34.5, 55.6]
top1_ic50 = [75.8, 17.5, 21.3, 29.0, 31.9, 31.1]

fig = go.Figure(data=[
    go.Bar(name='TOP3B IC50 (µM)', x=compounds, y=top3b_ic50, marker_color='#6A0C76'),
    go.Bar(name='TOP1 IC50 (µM)', x=compounds, y=top1_ic50, marker_color='#A020F0')
])

fig.update_layout(barmode='group', title='Inhibitory Potency Comparison', xaxis_title='Compound', yaxis_title='IC50 (µM)')
fig.show()
```

The above code provides a clear bar plot to compare the potencies, enabling visual identification of compounds with superior TOP3B selectivity.
