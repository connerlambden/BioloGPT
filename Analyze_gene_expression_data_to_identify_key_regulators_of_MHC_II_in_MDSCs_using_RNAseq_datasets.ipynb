This analysis will utilize RNA-seq data from MDSCs to identify differentially expressed genes associated with MHC II regulation.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load RNA-seq data
data = pd.read_csv('path_to_rnaseq_data.csv')

# Analyze expression levels of MHC II related genes
mhc_genes = ['HLA-DRA', 'HLA-DRB1', 'CIITA']
expression_data = data[mhc_genes]

# Visualize expression levels
sns.boxplot(data=expression_data)
plt.title('MHC II Gene Expression in MDSCs')
plt.show()
```

This code will help visualize the expression levels of MHC II related genes in MDSCs, providing insights into their regulation.
