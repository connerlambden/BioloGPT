This notebook will analyze nutrient uptake data from glacier ice algae, focusing on phosphorus levels and their correlation with algal biomass and albedo.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load nutrient uptake data
data = pd.read_csv('nutrient_uptake_data.csv')

# Analyze correlation between phosphorus levels and algal biomass
correlation = data['phosphorus'].corr(data['biomass'])

# Plotting the data
plt.scatter(data['phosphorus'], data['biomass'])
plt.title('Phosphorus Levels vs Algal Biomass')
plt.xlabel('Phosphorus Levels (mg/L)')
plt.ylabel('Algal Biomass (cells/ml)')
plt.show()
```

The analysis shows the correlation between phosphorus levels and algal biomass, providing insights into how nutrient availability affects algal growth.
