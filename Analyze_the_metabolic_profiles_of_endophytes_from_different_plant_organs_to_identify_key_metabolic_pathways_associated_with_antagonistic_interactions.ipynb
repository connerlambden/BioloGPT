This analysis will utilize datasets from the study of endophytes in *O. heracleoticum* to explore metabolic differences.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load metabolic data from the study
metabolic_data = pd.read_csv('metabolic_profiles.csv')

# Analyze metabolic activity by organ
activity_summary = metabolic_data.groupby('organ').mean()

# Visualize the results
activity_summary.plot(kind='bar')
plt.title('Average Metabolic Activity by Plant Organ')
plt.ylabel('Metabolic Activity')
plt.show()
```

This code will help visualize the differences in metabolic activity among endophytes isolated from different organs of *O. heracleoticum*.
