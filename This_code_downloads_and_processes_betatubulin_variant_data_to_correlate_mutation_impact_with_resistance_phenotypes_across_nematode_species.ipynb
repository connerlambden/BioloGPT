Retrieve genomic variant data for Caenorhabditis species and filter for beta-tubulin mutations (e.g., ben-1) to evaluate their association with benzimidazole resistance.

```python
import pandas as pd
# Load variant data
data = pd.read_csv('path/to/variant_data.csv')
# Filter for beta-tubulin gene mutations
ben1_variants = data[data['gene'] == 'ben-1']
# Group by species and calculate average impact score
impact_summary = ben1_variants.groupby('species')['impact_score'].mean()
print(impact_summary)
```

Visualize the association between beta-tubulin mutation impact and resistance phenotype using a bar plot.

```python
import matplotlib.pyplot as plt
impact_summary.plot(kind='bar', color='#6A0C76')
plt.title('Average Beta-Tubulin Mutation Impact by Species')
plt.xlabel('Species')
plt.ylabel('Average Impact Score')
plt.show()
```
