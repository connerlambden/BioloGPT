```python
import plotly.graph_objs as go

# Data for plotting
concentrations = ['0.5 mM', '1 mM', '5 mM']
yield_with = [0.3, 0.3, 0.3]
yield_without = [0.1, 0.1, 0.1]

trace1 = go.Bar(x=concentrations, y=yield_with, name='With Polyarginine', marker=dict(color='#6A0C76'))
trace2 = go.Bar(x=concentrations, y=yield_without, name='Without Polyarginine', marker=dict(color='#B4A7D6'))
data = [trace1, trace2]

layout = go.Layout(title='Relative NADH Yield Under Different Conditions', barmode='group', 
                   xaxis=dict(title='NADH Concentration'), yaxis=dict(title='Relative NADH Yield'))

fig = go.Figure(data=data, layout=layout)
fig.show()
```
