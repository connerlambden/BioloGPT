This notebook analyzes DNA methylation and RNA sequencing data from postmortem brain samples.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Download data from specified repository
methylation_data = pd.read_csv('https://www.ebi.ac.uk/arrayexpress/experiments/E-GEOD-42002')
expression_data = pd.read_csv('https://www.ebi.ac.uk/arrayexpress/experiments/E-GEOD-42902')
# Merge datasets and perform correlation analysis
merged = pd.merge(methylation_data, expression_data, on='sample_id')
correlation = merged['FKBP5_methylation'].corr(merged['FKBP5_expression'])
print('Correlation:', correlation)
plt.scatter(merged['FKBP5_methylation'], merged['FKBP5_expression'], color='blue')
plt.xlabel('FKBP5 Methylation')
plt.ylabel('FKBP5 Expression')
plt.title('Correlation between FKBP5 Methylation and Expression')
plt.show()
```

This code snippet performs a simple correlation analysis to demonstrate the inverse relationship between methylation and expression levels.

```python
# Further analysis could include stratification by age and diagnosis
age_stratified = merged.groupby('age_group').corr().loc[:, ['FKBP5_methylation', 'FKBP5_expression']]
print(age_stratified)
```
