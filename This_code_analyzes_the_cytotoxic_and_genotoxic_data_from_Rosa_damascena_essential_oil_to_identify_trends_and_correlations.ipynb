### Data Analysis of Rosa damascena Essential Oil Effects

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Concentration (µg/mL)': [50, 100, 500],
        'Chromosomal Aberrations (%)': [6.00, 6.00, 8.40]}
df = pd.DataFrame(data)

plt.bar(df['Concentration (µg/mL)'], df['Chromosomal Aberrations (%)'], color='purple')
plt.title('Chromosomal Aberrations in Human Lymphocytes')
plt.xlabel('Concentration (µg/mL)')
plt.ylabel('Chromosomal Aberrations (%)')
plt.show()
```

### Discussion
This analysis provides a visual representation of the cytotoxic effects of Rosa damascena essential oil, highlighting the concentration-dependent increase in chromosomal aberrations.
