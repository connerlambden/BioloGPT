### Data Analysis Overview
This notebook reads the athlete dataset and produces statistical comparisons between WHO BMI and BF% categorizations.

```python
import pandas as pd
import plotly.express as px

# Loading study dataset
data = {
    'Category': ['Normal Weight', 'Overweight', 'Obesity'],
    'WHO_BMI': [451, 148, 23],
    'BF_Percent': [598, 19, 5]
}
df = pd.DataFrame(data)

# Plotly Bar chart
fig = px.bar(df, x='Category', y=['WHO_BMI', 'BF_Percent'], barmode='group', 
             title='Comparison of WHO BMI vs BF% Classifications',
             labels={'value': 'Counts', 'variable': 'Method'})
fig.update_layout(title_font_color='#6A0C76')
fig.show()
```

### Interpretation
The graph illustrates the discrepancies between traditional WHO BMI classification and BF%-based categorization in athletes.
