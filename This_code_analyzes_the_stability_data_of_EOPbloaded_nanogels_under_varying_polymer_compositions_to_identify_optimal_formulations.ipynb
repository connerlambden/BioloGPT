Analyze stability data of EOPb-loaded nanogels.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load stability data
stability_data = pd.read_csv('EOPb_nanogel_stability.csv')

# Plot stability over time
plt.figure(figsize=(10,6))
plt.plot(stability_data['Time'], stability_data['Stability'], marker='o')
plt.title('Stability of EOPb-Loaded Nanogels Over Time')
plt.xlabel('Time (days)')
plt.ylabel('Stability (%)')
plt.grid()
plt.show()
```
