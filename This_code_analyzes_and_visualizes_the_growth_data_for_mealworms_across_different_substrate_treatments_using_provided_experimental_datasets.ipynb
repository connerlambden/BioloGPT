Load and inspect the dataset containing treatment groups, weight, yield, and mortality data for mealworms.

```python
import pandas as pd
import plotly.express as px

data = {'Treatment': ['100% wheat bran', '50% wheat bran + bean waste', '50% wheat bran + strawberry waste'],
        'Mean_Larva_Weight': [124, 135, 95],
        'Fresh_Yield': ['Highest', 'Not significantly lower', 'Lowest'],
        'Dry_Yield': ['Highest', 'Not significantly lower', 'Lowest'],
        'Mortality': ['Highest', 'Not significantly different', 'Lower than control']}
df = pd.DataFrame(data)

fig = px.bar(df, x='Treatment', y='Mean_Larva_Weight', color='Treatment', title='Mean Larva Weight Across Treatments')
fig.show()
```
