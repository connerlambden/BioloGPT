### Data Analysis of Adhesive Pull-Off Force
This notebook analyzes the effects of surface roughness and contact angle on adhesive pull-off force using provided datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for analysis
roughness_data = {'Surface Roughness (µm)': [0.1, 0.5, 1.0, 1.5, 2.0],
                  'Contact Angle (°)': [30, 45, 60, 75, 90],
                  'Adhesive Pull-Off Force (N)': [5, 10, 15, 12, 8]}

# Create DataFrame
roughness_df = pd.DataFrame(roughness_data)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.scatter(roughness_df['Surface Roughness (µm)'], roughness_df['Adhesive Pull-Off Force (N)'],
            c=roughness_df['Contact Angle (°)'], cmap='viridis')
plt.colorbar(label='Contact Angle (°)')
plt.title('Effect of Surface Roughness and Contact Angle on Adhesive Pull-Off Force')
plt.xlabel('Surface Roughness (µm)')
plt.ylabel('Adhesive Pull-Off Force (N)')
plt.grid(True)
plt.show()
```
