This code downloads genetic gain data and generates interactive Plotly graphs to visualize trends in BPI and ASI over time.

```python
import plotly.graph_objs as go
import plotly.offline as pyo

years = [2021, 2022, 2023, 2024, 2025, 2026]
bpi = [136, 161, 184, 312, 320, 384]
asi = [65, 73, 94, 142, 135, 172]

trace1 = go.Scatter(x=years, y=bpi, mode='lines+markers', name='BPI', line=dict(color='#6A0C76'))
trace2 = go.Scatter(x=years, y=asi, mode='lines+markers', name='ASI', line=dict(color='#FF5733'))

data = [trace1, trace2]

layout = go.Layout(title='Genetic Gain Trends in Dairy Herd', xaxis=dict(title='Year Born'), yaxis=dict(title='Index Value'))

fig = go.Figure(data=data, layout=layout)
pyo.plot(fig, filename='genetic_gain_trends.html')
```
