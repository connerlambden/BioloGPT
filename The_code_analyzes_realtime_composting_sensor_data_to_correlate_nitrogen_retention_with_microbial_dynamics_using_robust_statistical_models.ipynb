This notebook downloads experimental sensor data on oxygen, temperature, and carbon levels during composting and correlates these with measured nitrogen retention outcomes.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
data = pd.read_csv('composting_sensor_data.csv')
x = data[['temperature', 'oxygen', 'carbon']]
y = data['nitrogen_retention']
x = sm.add_constant(x)
model = sm.OLS(y, x).fit()
print(model.summary())
```

The above code performs linear regression analysis on the composting data to quantify how environmental factors predict nitrogen retention.

```python
# Plotting the correlation
import matplotlib.pyplot as plt
plt.scatter(data['temperature'], data['nitrogen_retention'], color='#6A0C76')
plt.xlabel('Temperature')
plt.ylabel('Nitrogen Retention (%)')
plt.title('Correlation between Temperature and Nitrogen Retention')
plt.show()
```

This visualization helps in understanding the degree of correlation between temperature and nitrogen retention in composting setups.
