### Data Analysis of Gata1 Mutations in Megakaryocytes

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load single-cell RNA-seq data
sc_data = pd.read_csv('gata1_mutation_data.csv')

# Analyze transcriptional changes
transcriptional_changes = sc_data.groupby('stage').mean()

# Visualize the results
plt.figure(figsize=(10, 6))
sns.heatmap(transcriptional_changes, annot=True, cmap='coolwarm')
plt.title('Transcriptional Changes in Megakaryocytes with Gata1 Mutations')
plt.show()
```
